## 2. ä½¿ç”¨create_react_agentåˆ›å»ºAgent

```python
# ä½¿ç”¨create_react_agentåˆ›å»ºAgent - ç‹¬ç«‹ä»£ç å—
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from langchain.agents import create_react_agent
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
import os

# åŠ è½½ç¯å¢ƒå˜é‡
load_dotenv()

print("ğŸ”§ ä½¿ç”¨create_react_agentåˆ›å»ºAgent:")
print("=" * 50)

openai_api_key = os.getenv('OPENAI_API_KEY')
if not openai_api_key:
    print("âŒ OpenAI API Key æœªé…ç½®")
else:
    try:
        # 1. å‡†å¤‡Calculatorå·¥å…·é›†åˆ
        print(f"ğŸ› ï¸  1. å‡†å¤‡Calculatorå·¥å…·é›†åˆ:")
        
        @tool
        def åŠ æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å’Œï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
            result = a + b
            return f"{a} + {b} = {result}"
        
        @tool
        def å‡æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å·®ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
            result = a - b
            return f"{a} - {b} = {result}"
        
        @tool
        def ä¹˜æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„ç§¯ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
            result = a * b
            return f"{a} Ã— {b} = {result}"
        
        @tool
        def é™¤æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å•†ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—ã€‚é™¤æ•°ä¸èƒ½ä¸ºé›¶"""
            if b == 0:
                return "é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸ºé›¶"
            result = a / b
            return f"{a} Ã· {b} = {result}"
        
        calculator_tools = [åŠ æ³•, å‡æ³•, ä¹˜æ³•, é™¤æ³•]
        
        print(f"   Calculatorå·¥å…·é›†åˆ: {len(calculator_tools)} ä¸ªå·¥å…·")
        for tool_obj in calculator_tools:
            print(f"     - {tool_obj.name}: {tool_obj.description}")
        
        # 2. åˆ›å»ºGPT LLM
        print(f"\nğŸ¤– 2. åˆ›å»ºGPT LLM:")
        
        llm = ChatOpenAI(
            model="gpt-4o-mini",
            temperature=0.1,
            max_tokens=200
        )
        
        print(f"   LLMé…ç½®: {llm.model_name}, temperature={llm.temperature}")
        
        # 3. åˆ›å»ºAgentæç¤ºæ¨¡æ¿
        print(f"\nğŸ“‹ 3. åˆ›å»ºAgentæç¤ºæ¨¡æ¿:")
        
        agent_prompt = PromptTemplate.from_template("""
ä½ æ˜¯ä¸€ä¸ªæ•°å­¦è®¡ç®—åŠ©æ‰‹ï¼Œå¯ä»¥ä½¿ç”¨ä»¥ä¸‹å·¥å…·æ¥è§£å†³æ•°å­¦é—®é¢˜ï¼š

å¯ç”¨å·¥å…·:
{tools}

å·¥å…·åç§°:
{tool_names}

è¯·ä½¿ç”¨ä»¥ä¸‹æ ¼å¼å›ç­”é—®é¢˜ï¼š

Question: ç”¨æˆ·çš„é—®é¢˜
Thought: æˆ‘éœ€è¦æ€è€ƒå¦‚ä½•è§£å†³è¿™ä¸ªé—®é¢˜
Action: æˆ‘éœ€è¦ä½¿ç”¨çš„å·¥å…·åç§°
Action Input: å·¥å…·çš„è¾“å…¥å‚æ•°
Observation: å·¥å…·æ‰§è¡Œçš„ç»“æœ
... (å¯ä»¥é‡å¤Thought/Action/Action Input/Observationå¤šæ¬¡)
Thought: æˆ‘ç°åœ¨çŸ¥é“æœ€ç»ˆç­”æ¡ˆäº†
Final Answer: æœ€ç»ˆçš„ç­”æ¡ˆ

å¼€å§‹ï¼

Question: {input}
Thought:{agent_scratchpad}
""")
        
        print(f"   Agentæç¤ºæ¨¡æ¿åˆ›å»ºå®Œæˆ")
        print(f"   æ¨¡æ¿å˜é‡: tools, tool_names, input, agent_scratchpad")
        
        # 4. ä½¿ç”¨create_react_agentåˆ›å»ºAgent
        print(f"\nğŸ—ï¸  4. ä½¿ç”¨create_react_agentåˆ›å»ºAgent:")
        
        print(f"   è°ƒç”¨ create_react_agent(llm, tools, prompt)...")
        
        agent = create_react_agent(
            llm=llm,
            tools=calculator_tools,
            prompt=agent_prompt
        )
        
        print(f"   Agentåˆ›å»ºæˆåŠŸ!")
        print(f"   Agentç±»å‹: {type(agent)}")
        print(f"   AgentåŒ…å«çš„LLM: {type(agent.llm)}")
        print(f"   AgentåŒ…å«çš„å·¥å…·: {len(agent.tools)} ä¸ª")
        
        # 5. éªŒè¯Agentç»„ä»¶
        print(f"\nğŸ” 5. éªŒè¯Agentç»„ä»¶:")
        
        print(f"   Agentç»„ä»¶æ£€æŸ¥:")
        print(f"     æœ‰LLM: {hasattr(agent, 'llm') and agent.llm is not None}")
        print(f"     æœ‰å·¥å…·: {hasattr(agent, 'tools') and len(agent.tools) > 0}")
        print(f"     æœ‰æç¤º: {hasattr(agent, 'prompt') and agent.prompt is not None}")
        print(f"     å¯è°ƒç”¨: {hasattr(agent, 'invoke')}")
        
        # æ˜¾ç¤ºå·¥å…·ä¿¡æ¯
        print(f"\n   Agentä¸­çš„å·¥å…·:")
        for i, tool_obj in enumerate(agent.tools, 1):
            print(f"     {i}. {tool_obj.name}: {tool_obj.description}")
        
        # 6. æµ‹è¯•AgentåŸºç¡€åŠŸèƒ½
        print(f"\nğŸ§ª 6. æµ‹è¯•AgentåŸºç¡€åŠŸèƒ½:")
        
        # åˆ›å»ºä¸€ä¸ªç®€å•çš„æµ‹è¯•è¾“å…¥
        test_input = "è¯·è®¡ç®— 15 + 25"
        
        print(f"   æµ‹è¯•è¾“å…¥: {test_input}")
        print(f"   è°ƒç”¨Agent.invoke()...")
        
        try:
            # ç›´æ¥è°ƒç”¨Agentï¼ˆä¸é€šè¿‡Executorï¼‰
            agent_response = agent.invoke({
                "input": test_input
            })
            
            print(f"   Agentå“åº”ç±»å‹: {type(agent_response)}")
            print(f"   Agentå“åº”å†…å®¹: {str(agent_response)[:200]}...")
            
        except Exception as e:
            print(f"   Agentç›´æ¥è°ƒç”¨é”™è¯¯: {e}")
            print(f"   è¯´æ˜ï¼šAgentéœ€è¦é€šè¿‡AgentExecutoræ‰§è¡Œ")
        
        # 7. Agentåˆ›å»ºæµç¨‹æ€»ç»“
        print(f"\nğŸ“ 7. Agentåˆ›å»ºæµç¨‹æ€»ç»“:")
        
        print(f"   create_react_agentè°ƒç”¨æˆåŠŸ:")
        print(f"     âœ“ å‚æ•°1: GPT LLM ({llm.model_name})")
        print(f"     âœ“ å‚æ•°2: å·¥å…·é›†åˆ ({len(calculator_tools)} ä¸ªå·¥å…·)")
        print(f"     âœ“ å‚æ•°3: æç¤ºæ¨¡æ¿ (ReActæ ¼å¼)")
        print(f"     âœ“ è¿”å›: å¯æ‰§è¡Œçš„Agentå¯¹è±¡")
        
        # ä¿å­˜Agentä¾›åç»­ä½¿ç”¨
        globals().update({
            'calculator_agent': agent,
            'calculator_tools': calculator_tools,
            'calculator_llm': llm,
            'calculator_prompt': agent_prompt
        })
        
        # éªŒè¯ç‚¹ï¼šcreate_react_agentåˆ›å»ºAgentæ­£ç¡®
        assert agent is not None, "Agentåº”è¯¥åˆ›å»ºæˆåŠŸ"
        assert hasattr(agent, 'llm'), "Agentåº”è¯¥æœ‰LLMå±æ€§"
        assert hasattr(agent, 'tools'), "Agentåº”è¯¥æœ‰toolså±æ€§"
        assert len(agent.tools) == 4, "Agentåº”è¯¥æœ‰4ä¸ªå·¥å…·"
        assert hasattr(agent, 'prompt'), "Agentåº”è¯¥æœ‰promptå±æ€§"
        
        print(f"\nâœ… éªŒè¯é€šè¿‡ï¼šcreate_react_agentåˆ›å»ºAgentæ­£ç¡®")
        print(f"\nğŸ¯ Agentåˆ›å»ºæ€»ç»“:")
        print(f"   âœ“ ä½¿ç”¨create_react_agent(GPT llm, tools, prompt)åˆ›å»ºAgent")
        print(f"   âœ“ é›†æˆäº†4ä¸ªCalculatorå·¥å…·")
        print(f"   âœ“ é…ç½®äº†ReActæ€è€ƒæ¨¡å¼")
        print(f"   âœ“ Agentå¯¹è±¡åˆ›å»ºæˆåŠŸ")
        print(f"   âœ“ å‡†å¤‡é…ç½®AgentExecutoræ‰§è¡Œå™¨")
        
    except Exception as e:
        print(f"\nâŒ ä½¿ç”¨create_react_agentåˆ›å»ºAgentå¤±è´¥: {e}")
```

```python
# éªŒè¯GPT Agentèƒ½è°ƒç”¨å·¥å…·è§£å†³é—®é¢˜ - ç‹¬ç«‹ä»£ç å—
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from langchain.agents import create_react_agent, AgentExecutor
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
import os
import re

# åŠ è½½ç¯å¢ƒå˜é‡
load_dotenv()

print("ğŸ” éªŒè¯GPT Agentèƒ½è°ƒç”¨å·¥å…·è§£å†³é—®é¢˜:")
print("=" * 50)

openai_api_key = os.getenv('OPENAI_API_KEY')
if not openai_api_key:
    print("âŒ OpenAI API Key æœªé…ç½®")
else:
    try:
        # 1. ä½¿ç”¨å·²é…ç½®çš„AgentExecutor
        print(f"ğŸ”„ 1. ä½¿ç”¨å·²é…ç½®çš„AgentExecutor:")
        
        # æ£€æŸ¥æ˜¯å¦æœ‰å·²ä¿å­˜çš„Executor
        if 'calculator_executor' in globals():
            executor = globals()['calculator_executor']
            print(f"   ä½¿ç”¨å·²ä¿å­˜çš„Executor: {type(executor)}")
        else:
            print(f"   é‡æ–°é…ç½®AgentExecutor...")
            
            # é‡æ–°åˆ›å»ºå®Œæ•´çš„Agentç³»ç»Ÿ
            @tool
            def åŠ æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å’Œï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
                result = a + b
                return f"{a} + {b} = {result}"
            
            @tool
            def å‡æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å·®ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
                result = a - b
                return f"{a} - {b} = {result}"
            
            @tool
            def ä¹˜æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„ç§¯ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
                result = a * b
                return f"{a} Ã— {b} = {result}"
            
            @tool
            def é™¤æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å•†ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—ã€‚é™¤æ•°ä¸èƒ½ä¸ºé›¶"""
                if b == 0:
                    return "é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸ºé›¶"
                result = a / b
                return f"{a} Ã· {b} = {result}"
            
            tools = [åŠ æ³•, å‡æ³•, ä¹˜æ³•, é™¤æ³•]
            llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.1, max_tokens=200)
            
            agent_prompt = PromptTemplate.from_template("""
ä½ æ˜¯ä¸€ä¸ªæ•°å­¦è®¡ç®—åŠ©æ‰‹ï¼Œå¯ä»¥ä½¿ç”¨ä»¥ä¸‹å·¥å…·æ¥è§£å†³æ•°å­¦é—®é¢˜ï¼š

å¯ç”¨å·¥å…·:
{tools}

å·¥å…·åç§°:
{tool_names}

è¯·ä½¿ç”¨ä»¥ä¸‹æ ¼å¼å›ç­”é—®é¢˜ï¼š

Question: ç”¨æˆ·çš„é—®é¢˜
Thought: æˆ‘éœ€è¦æ€è€ƒå¦‚ä½•è§£å†³è¿™ä¸ªé—®é¢˜
Action: æˆ‘éœ€è¦ä½¿ç”¨çš„å·¥å…·åç§°
Action Input: å·¥å…·çš„è¾“å…¥å‚æ•°
Observation: å·¥å…·æ‰§è¡Œçš„ç»“æœ
... (å¯ä»¥é‡å¤Thought/Action/Action Input/Observationå¤šæ¬¡)
Thought: æˆ‘ç°åœ¨çŸ¥é“æœ€ç»ˆç­”æ¡ˆäº†
Final Answer: æœ€ç»ˆçš„ç­”æ¡ˆ

å¼€å§‹ï¼

Question: {input}
Thought:{agent_scratchpad}
""")
            
            agent = create_react_agent(llm, tools, agent_prompt)
            executor = AgentExecutor(
                agent=agent,
                tools=tools,
                verbose=True,
                max_iterations=5,
                early_stopping_method="generate"
            )
            print(f"   AgentExecutoré‡æ–°é…ç½®å®Œæˆ")
        
        # 2. è®¾è®¡å¤æ‚é—®é¢˜æµ‹è¯•
        print(f"\nğŸ§® 2. è®¾è®¡å¤æ‚é—®é¢˜æµ‹è¯•:")
        
        complex_problems = [
            {
                "question": "è¯·è®¡ç®— 25 + 17ï¼Œç„¶åå°†ç»“æœä¹˜ä»¥ 3",
                "expected_tools": ["åŠ æ³•", "ä¹˜æ³•"],
                "description": "å¤šæ­¥éª¤è®¡ç®—é—®é¢˜"
            },
            {
                "question": "è¯·è®¡ç®— (100 - 20) Ã· 4",
                "expected_tools": ["å‡æ³•", "é™¤æ³•"],
                "description": "æ‹¬å·è¿ç®—é—®é¢˜"
            },
            {
                "question": "è¯·è®¡ç®— 15 Ã— 6 + 10",
                "expected_tools": ["ä¹˜æ³•", "åŠ æ³•"],
                "description": "æ··åˆè¿ç®—é—®é¢˜"
            },
            {
                "question": "è¯·è®¡ç®— 50 Ã· (2 + 3)",
                "expected_tools": ["åŠ æ³•", "é™¤æ³•"],
                "description": "å¤æ‚åˆ†å¼é—®é¢˜"
            }
        ]
        
        print(f"   è®¾è®¡äº† {len(complex_problems)} ä¸ªå¤æ‚é—®é¢˜:")
        for i, problem in enumerate(complex_problems, 1):
            print(f"     {i}. {problem['description']}: {problem['question']}")
            print(f"        é¢„æœŸå·¥å…·: {', '.join(problem['expected_tools'])}")
        
        # 3. æ‰§è¡Œå¤æ‚é—®é¢˜æµ‹è¯•
        print(f"\nğŸš€ 3. æ‰§è¡Œå¤æ‚é—®é¢˜æµ‹è¯•:")
        
        problem_results = []
        
        for i, problem in enumerate(complex_problems, 1):
            print(f"\n   ğŸ“ é—®é¢˜ {i}: {problem['description']}")
            print(f"      é—®é¢˜: {problem['question']}")
            
            try:
                # æ‰§è¡ŒAgent
                result = executor.invoke({
                    "input": problem['question']
                })
                
                output = result.get('output', '')
                print(f"\n      ğŸ’¬ Agentå›ç­”: {output}")
                
                # åˆ†æå·¥å…·ä½¿ç”¨æƒ…å†µ
                tools_used = []
                for tool_name in problem['expected_tools']:
                    if tool_name in output:
                        tools_used.append(tool_name)
                
                problem_results.append({
                    "problem": problem['question'],
                    "description": problem['description'],
                    "expected_tools": problem['expected_tools'],
                    "agent_output": output,
                    "tools_detected": tools_used,
                    "success": len(output) > 10  # ç®€å•æˆåŠŸåˆ¤æ–­
                })
                
                print(f"      âœ… æ‰§è¡ŒæˆåŠŸ")
                
            except Exception as e:
                print(f"      âŒ æ‰§è¡Œå¤±è´¥: {e}")
                problem_results.append({
                    "problem": problem['question'],
                    "description": problem['description'],
                    "success": False,
                    "error": str(e)
                })
        
        # 4. å·¥å…·è°ƒç”¨éªŒè¯
        print(f"\nğŸ”§ 4. å·¥å…·è°ƒç”¨éªŒè¯:")
        
        def verify_tool_usage(problem_result):
            """éªŒè¯å·¥å…·ä½¿ç”¨æƒ…å†µ"""
            if not problem_result.get("success", False):
                return {"overall_success": False, "error": "æ‰§è¡Œå¤±è´¥"}
            
            agent_output = problem_result.get("agent_output", "")
            expected_tools = problem_result.get("expected_tools", [])
            detected_tools = problem_result.get("tools_detected", [])
            
            verification = {
                "has_output": len(agent_output) > 0,
                "tools_called": len(detected_tools) > 0,
                "expected_tools_used": len(detected_tools) >= len(expected_tools) * 0.5,  # è‡³å°‘ä½¿ç”¨ä¸€åŠé¢„æœŸå·¥å…·
                "output_meaningful": len(agent_output) > 10,
                "contains_numbers": bool(re.search(r'\d+', agent_output))
            }
            
            verification["overall_success"] = all(verification.values())
            return verification
        
        print(f"   å·¥å…·è°ƒç”¨éªŒè¯ç»“æœ:")
        verification_summary = []
        
        for i, result in enumerate(problem_results, 1):
            print(f"\n     é—®é¢˜ {i}: {result['description']}")
            
            verification = verify_tool_usage(result)
            verification_summary.append(verification)
            
            print(f"       éªŒè¯ç»“æœ:")
            for check_name, passed in verification.items():
                if check_name == "error":
                    print(f"         {check_name}: {passed}")
                elif check_name == "overall_success":
                    status = "âœ…" if passed else "âŒ"
                    print(f"         {check_name}: {status}")
                else:
                    status = "âœ…" if passed else "âŒ"
                    print(f"         {check_name}: {status}")
        
        # 5. è‡ªåŠ¨å·¥å…·é€‰æ‹©éªŒè¯
        print(f"\nğŸ¤– 5. è‡ªåŠ¨å·¥å…·é€‰æ‹©éªŒè¯:")
        
        # ç»Ÿè®¡å·¥å…·ä½¿ç”¨æƒ…å†µ
        tool_usage_stats = {}
        for result in problem_results:
            if result.get("success", False):
                detected_tools = result.get("tools_detected", [])
                for tool in detected_tools:
                    tool_usage_stats[tool] = tool_usage_stats.get(tool, 0) + 1
        
        print(f"   å·¥å…·ä½¿ç”¨ç»Ÿè®¡:")
        for tool_name, count in tool_usage_stats.items():
            print(f"     {tool_name}: {count} æ¬¡")
        
        # éªŒè¯è‡ªåŠ¨é€‰æ‹©èƒ½åŠ›
        auto_selection_success = len(tool_usage_stats) > 0
        print(f"\n   è‡ªåŠ¨å·¥å…·é€‰æ‹©: {'âœ… æˆåŠŸ' if auto_selection_success else 'âŒ å¤±è´¥'}")
        
        # 6. é—®é¢˜è§£å†³èƒ½åŠ›éªŒè¯
        print(f"\nğŸ¯ 6. é—®é¢˜è§£å†³èƒ½åŠ›éªŒè¯:")
        
        successful_problems = sum(1 for r in problem_results if r.get("success", False))
        total_problems = len(problem_results)
        problem_success_rate = successful_problems / total_problems * 100 if total_problems > 0 else 0
        
        successful_verifications = sum(1 for v in verification_summary if v.get("overall_success", False))
        verification_success_rate = successful_verifications / len(verification_summary) * 100 if verification_summary else 0
        
        print(f"   é—®é¢˜è§£å†³ç»Ÿè®¡:")
        print(f"     æ€»é—®é¢˜æ•°: {total_problems}")
        print(f"     è§£å†³æˆåŠŸæ•°: {successful_problems}")
        print(f"     è§£å†³æˆåŠŸç‡: {problem_success_rate:.1f}%")
        print(f"\n   éªŒè¯é€šè¿‡ç»Ÿè®¡:")
        print(f"     éªŒè¯æˆåŠŸæ•°: {successful_verifications}")
        print(f"     éªŒè¯æˆåŠŸç‡: {verification_success_rate:.1f}%")
        
        # 7. ä¿å­˜æµ‹è¯•ç»“æœ
        print(f"\nğŸ’¾ 7. ä¿å­˜æµ‹è¯•ç»“æœ:")
        
        globals().update({
            'agent_problem_results': problem_results,
            'agent_verification_summary': verification_summary,
            'tool_usage_stats': tool_usage_stats
        })
        
        print(f"   æµ‹è¯•ç»“æœå·²ä¿å­˜åˆ°å…¨å±€å˜é‡")
        
        # éªŒè¯ç‚¹ï¼šGPT Agentèƒ½è°ƒç”¨å·¥å…·è§£å†³é—®é¢˜
        assert total_problems == 4, "åº”è¯¥æœ‰4ä¸ªæµ‹è¯•é—®é¢˜"
        assert successful_problems >= 2, "è‡³å°‘åº”è¯¥æœ‰2ä¸ªé—®é¢˜è§£å†³æˆåŠŸ"
        assert problem_success_rate >= 50, "é—®é¢˜è§£å†³æˆåŠŸç‡åº”è¯¥è‡³å°‘50%"
        assert auto_selection_success, "åº”è¯¥èƒ½å¤Ÿè‡ªåŠ¨é€‰æ‹©å·¥å…·"
        assert len(tool_usage_stats) > 0, "åº”è¯¥æœ‰å·¥å…·ä½¿ç”¨ç»Ÿè®¡"
        
        print(f"\nâœ… éªŒè¯é€šè¿‡ï¼šGPT Agentèƒ½è°ƒç”¨å·¥å…·è§£å†³é—®é¢˜")
        print(f"\nğŸ¯ Agentå·¥å…·è°ƒç”¨éªŒè¯æ€»ç»“:")
        print(f"   âœ“ GPT Agentèƒ½å¤Ÿåˆ†æå¤æ‚é—®é¢˜")
        print(f"   âœ“ GPT Agentèƒ½å¤Ÿè‡ªåŠ¨é€‰æ‹©åˆé€‚çš„å·¥å…·")
        print(f"   âœ“ GPT Agentèƒ½å¤Ÿæ‰§è¡Œå¤šæ­¥éª¤è®¡ç®—")
        print(f"   âœ“ GPT Agentèƒ½å¤Ÿå¤„ç†æ··åˆè¿ç®—")
        print(f"   âœ“ å·¥å…·è°ƒç”¨æœºåˆ¶éªŒè¯æˆåŠŸ")
        
    except Exception as e:
        print(f"\nâŒ éªŒè¯GPT Agentèƒ½è°ƒç”¨å·¥å…·è§£å†³é—®é¢˜å¤±è´¥: {e}")
```

## 5. éªŒè¯GPT Agentèƒ½è‡ªåŠ¨é€‰æ‹©å¹¶ä½¿ç”¨å·¥å…·

```python
# éªŒè¯GPT Agentèƒ½è‡ªåŠ¨é€‰æ‹©å¹¶ä½¿ç”¨å·¥å…· - ç‹¬ç«‹ä»£ç å—
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from langchain.agents import create_react_agent, AgentExecutor
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
import os
import re
import json

# åŠ è½½ç¯å¢ƒå˜é‡
load_dotenv()

print("ğŸ¯ éªŒè¯GPT Agentèƒ½è‡ªåŠ¨é€‰æ‹©å¹¶ä½¿ç”¨å·¥å…·:")
print("=" * 50)

openai_api_key = os.getenv('OPENAI_API_KEY')
if not openai_api_key:
    print("âŒ OpenAI API Key æœªé…ç½®")
else:
    try:
        # 1. åˆ›å»ºå¢å¼ºçš„å·¥å…·é›†åˆ
        print(f"ğŸ› ï¸  1. åˆ›å»ºå¢å¼ºçš„å·¥å…·é›†åˆ:")
        
        @tool
        def åŠ æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å’Œï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
            result = a + b
            return f"{a} + {b} = {result}"
        
        @tool
        def å‡æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å·®ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
            result = a - b
            return f"{a} - {b} = {result}"
        
        @tool
        def ä¹˜æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„ç§¯ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
            result = a * b
            return f"{a} Ã— {b} = {result}"
        
        @tool
        def é™¤æ³•(a: float, b: float) -> str:
            """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å•†ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—ã€‚é™¤æ•°ä¸èƒ½ä¸ºé›¶"""
            if b == 0:
                return "é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸ºé›¶"
            result = a / b
            return f"{a} Ã· {b} = {result}"
        
        @tool
        def å¹³æ–¹æ ¹(x: float) -> str:
            """è®¡ç®—ä¸€ä¸ªéè´Ÿæ•°çš„å¹³æ–¹æ ¹"""
            import math
            if x < 0:
                return "é”™è¯¯ï¼šä¸èƒ½è®¡ç®—è´Ÿæ•°çš„å¹³æ–¹æ ¹"
            result = math.sqrt(x)
            return f"âˆš{x} = {result}"
        
        @tool
        def å¹‚è¿ç®—(base: float, exponent: float) -> str:
            """è®¡ç®—baseçš„exponentæ¬¡å¹‚"""
            result = base ** exponent
            return f"{base}^{exponent} = {result}"
        
        enhanced_tools = [åŠ æ³•, å‡æ³•, ä¹˜æ³•, é™¤æ³•, å¹³æ–¹æ ¹, å¹‚è¿ç®—]
        
        print(f"   å¢å¼ºå·¥å…·é›†åˆ: {len(enhanced_tools)} ä¸ªå·¥å…·")
        for tool_obj in enhanced_tools:
            print(f"     - {tool_obj.name}: {tool_obj.description}")
        
        # 2. åˆ›å»ºæ™ºèƒ½Agent
        print(f"\nğŸ¤– 2. åˆ›å»ºæ™ºèƒ½Agent:")
        
        llm = ChatOpenAI(
            model="gpt-4o-mini",
            temperature=0.1,
            max_tokens=300
        )
        
        # å¢å¼ºçš„æç¤ºæ¨¡æ¿
        enhanced_prompt = PromptTemplate.from_template("""
ä½ æ˜¯ä¸€ä¸ªä¸“ä¸šçš„æ•°å­¦è®¡ç®—åŠ©æ‰‹ï¼Œå¯ä»¥ä½¿ç”¨ä»¥ä¸‹å·¥å…·æ¥è§£å†³å„ç§æ•°å­¦é—®é¢˜ï¼š

å¯ç”¨å·¥å…·:
{tools}

å·¥å…·åç§°:
{tool_names}

ä½¿ç”¨æŒ‡å—:
1. ä»”ç»†åˆ†æç”¨æˆ·çš„æ•°å­¦é—®é¢˜
2. æ ¹æ®é—®é¢˜ç±»å‹é€‰æ‹©æœ€åˆé€‚çš„å·¥å…·
3. å¯¹äºå¤æ‚é—®é¢˜ï¼Œå¯ä»¥åˆ†æ­¥ä½¿ç”¨å¤šä¸ªå·¥å…·
4. ç¡®ä¿è®¡ç®—é¡ºåºæ­£ç¡®ï¼ˆå…ˆä¹˜é™¤ååŠ å‡ï¼‰
5. æ£€æŸ¥ç‰¹æ®Šæƒ…å†µï¼ˆå¦‚é™¤é›¶ã€è´Ÿæ•°å¹³æ–¹æ ¹ç­‰ï¼‰

è¯·ä½¿ç”¨ä»¥ä¸‹æ ¼å¼å›ç­”é—®é¢˜ï¼š

Question: ç”¨æˆ·çš„é—®é¢˜
Thought: æˆ‘éœ€è¦åˆ†æè¿™ä¸ªé—®é¢˜å¹¶é€‰æ‹©åˆé€‚çš„å·¥å…·
Action: æˆ‘éœ€è¦ä½¿ç”¨çš„å·¥å…·åç§°
Action Input: å·¥å…·çš„è¾“å…¥å‚æ•°
Observation: å·¥å…·æ‰§è¡Œçš„ç»“æœ
... (å¯ä»¥é‡å¤Thought/Action/Action Input/Observationå¤šæ¬¡)
Thought: æˆ‘ç°åœ¨çŸ¥é“æœ€ç»ˆç­”æ¡ˆäº†
Final Answer: æœ€ç»ˆçš„ç­”æ¡ˆ

å¼€å§‹ï¼

Question: {input}
Thought:{agent_scratchpad}
""")
        
        agent = create_react_agent(llm, enhanced_tools, enhanced_prompt)
        
        executor = AgentExecutor(
            agent=agent,
            tools=enhanced_tools,
            verbose=True,
            max_iterations=6,
            early_stopping_method="generate"
        )
        
        print(f"   æ™ºèƒ½Agentåˆ›å»ºå®Œæˆ")
        print(f"   å·¥å…·æ•°é‡: {len(enhanced_tools)}")
        print(f"   æœ€å¤§è¿­ä»£: {executor.max_iterations}")
        
        # 3. è®¾è®¡å·¥å…·é€‰æ‹©æµ‹è¯•æ¡ˆä¾‹
        print(f"\nğŸ§® 3. è®¾è®¡å·¥å…·é€‰æ‹©æµ‹è¯•æ¡ˆä¾‹:")
        
        selection_test_cases = [
            {
                "question": "è¯·è®¡ç®— 25 + 17",
                "expected_tool": "åŠ æ³•",
                "difficulty": "ç®€å•",
                "operation_type": "åŸºç¡€è¿ç®—"
            },
            {
                "question": "è¯·è®¡ç®— 8 Ã— 9",
                "expected_tool": "ä¹˜æ³•",
                "difficulty": "ç®€å•",
                "operation_type": "åŸºç¡€è¿ç®—"
            },
            {
                "question": "è¯·è®¡ç®— 144 çš„å¹³æ–¹æ ¹",
                "expected_tool": "å¹³æ–¹æ ¹",
                "difficulty": "ä¸­ç­‰",
                "operation_type": "é«˜çº§è¿ç®—"
            },
            {
                "question": "è¯·è®¡ç®— 2 çš„ 10 æ¬¡å¹‚",
                "expected_tool": "å¹‚è¿ç®—",
                "difficulty": "ä¸­ç­‰",
                "operation_type": "é«˜çº§è¿ç®—"
            },
            {
                "question": "è¯·è®¡ç®— (15 + 25) Ã— 3",
                "expected_tools": ["åŠ æ³•", "ä¹˜æ³•"],
                "difficulty": "å¤æ‚",
                "operation_type": "æ··åˆè¿ç®—"
            },
            {
                "question": "è¯·è®¡ç®— 100 Ã· (25 - 15)",
                "expected_tools": ["å‡æ³•", "é™¤æ³•"],
                "difficulty": "å¤æ‚",
                "operation_type": "æ··åˆè¿ç®—"
            }
        ]
        
        print(f"   è®¾è®¡äº† {len(selection_test_cases)} ä¸ªå·¥å…·é€‰æ‹©æµ‹è¯•:")
        for i, case in enumerate(selection_test_cases, 1):
            expected = case['expected_tool'] if isinstance(case['expected_tool'], str) else ', '.join(case['expected_tools'])
            print(f"     {i}. [{case['difficulty']}] {case['question']}")
            print(f"        é¢„æœŸå·¥å…·: {expected} ({case['operation_type']})")
        
        # 4. æ‰§è¡Œå·¥å…·é€‰æ‹©æµ‹è¯•
        print(f"\nğŸš€ 4. æ‰§è¡Œå·¥å…·é€‰æ‹©æµ‹è¯•:")
        
        selection_results = []
        
        for i, test_case in enumerate(selection_test_cases, 1):
            print(f"\n   ğŸ“ æµ‹è¯• {i}: {test_case['question']}")
            print(f"      éš¾åº¦: {test_case['difficulty']}, ç±»å‹: {test_case['operation_type']}")
            
            try:
                # æ‰§è¡ŒAgent
                result = executor.invoke({
                    "input": test_case['question']
                })
                
                output = result.get('output', '')
                print(f"\n      ğŸ’¬ Agentå›ç­”: {output}")
                
                # åˆ†æå·¥å…·é€‰æ‹©æƒ…å†µ
                expected_tools = test_case['expected_tool']
                if isinstance(expected_tools, str):
                    expected_tools = [expected_tools]
                
                detected_tools = []
                for tool_name in expected_tools:
                    if tool_name in output:
                        detected_tools.append(tool_name)
                
                # æ£€æŸ¥æ˜¯å¦ä½¿ç”¨äº†é¢„æœŸå·¥å…·
                correct_selection = len(detected_tools) >= len(expected_tools) * 0.5
                
                selection_results.append({
                    "question": test_case['question'],
                    "difficulty": test_case['difficulty'],
                    "expected_tools": expected_tools,
                    "detected_tools": detected_tools,
                    "agent_output": output,
                    "correct_selection": correct_selection,
                    "success": len(output) > 10
                })
                
                status = "âœ… æ­£ç¡®" if correct_selection else "âŒ é”™è¯¯"
                print(f"      å·¥å…·é€‰æ‹©: {status}")
                
            except Exception as e:
                print(f"      âŒ æ‰§è¡Œå¤±è´¥: {e}")
                selection_results.append({
                    "question": test_case['question'],
                    "success": False,
                    "error": str(e)
                })
        
        # 5. è‡ªåŠ¨é€‰æ‹©èƒ½åŠ›åˆ†æ
        print(f"\nğŸ“Š 5. è‡ªåŠ¨é€‰æ‹©èƒ½åŠ›åˆ†æ:")
        
        # ç»Ÿè®¡å„éš¾åº¦çº§åˆ«çš„è¡¨ç°
        difficulty_stats = {}
        for result in selection_results:
            if result.get("success", False):
                difficulty = result.get("difficulty", "æœªçŸ¥")
                if difficulty not in difficulty_stats:
                    difficulty_stats[difficulty] = {"total": 0, "correct": 0}
                difficulty_stats[difficulty]["total"] += 1
                if result.get("correct_selection", False):
                    difficulty_stats[difficulty]["correct"] += 1
        
        print(f"   æŒ‰éš¾åº¦çº§åˆ«åˆ†æ:")
        for difficulty, stats in difficulty_stats.items():
            accuracy = stats["correct"] / stats["total"] * 100 if stats["total"] > 0 else 0
            print(f"     {difficulty}: {stats['correct']}/{stats['total']} ({accuracy:.1f}%)")
        
        # 6. å·¥å…·é€‰æ‹©å‡†ç¡®æ€§éªŒè¯
        print(f"\nğŸ¯ 6. å·¥å…·é€‰æ‹©å‡†ç¡®æ€§éªŒè¯:")
        
        total_tests = len(selection_results)
        successful_tests = sum(1 for r in selection_results if r.get("success", False))
        correct_selections = sum(1 for r in selection_results if r.get("correct_selection", False))
        
        overall_success_rate = successful_tests / total_tests * 100 if total_tests > 0 else 0
        selection_accuracy = correct_selections / total_tests * 100 if total_tests > 0 else 0
        
        print(f"   æ•´ä½“ç»Ÿè®¡:")
        print(f"     æ€»æµ‹è¯•æ•°: {total_tests}")
        print(f"     æ‰§è¡ŒæˆåŠŸæ•°: {successful_tests}")
        print(f"     å·¥å…·é€‰æ‹©æ­£ç¡®æ•°: {correct_selections}")
        print(f"     æ‰§è¡ŒæˆåŠŸç‡: {overall_success_rate:.1f}%")
        print(f"     é€‰æ‹©å‡†ç¡®ç‡: {selection_accuracy:.1f}%")
        
        # 7. éªŒè¯ç‚¹æ£€æŸ¥
        print(f"\nâœ… 7. éªŒè¯ç‚¹æ£€æŸ¥:")
        
        verification_points = [
            (total_tests == 6, "åº”è¯¥æœ‰6ä¸ªæµ‹è¯•æ¡ˆä¾‹"),
            (successful_tests >= 4, "è‡³å°‘åº”è¯¥æœ‰4ä¸ªæµ‹è¯•æ‰§è¡ŒæˆåŠŸ"),
            (overall_success_rate >= 66.7, "æ‰§è¡ŒæˆåŠŸç‡åº”è¯¥è‡³å°‘66.7%"),
            (correct_selections >= 3, "è‡³å°‘åº”è¯¥æœ‰3ä¸ªå·¥å…·é€‰æ‹©æ­£ç¡®"),
            (selection_accuracy >= 50, "å·¥å…·é€‰æ‹©å‡†ç¡®ç‡åº”è¯¥è‡³å°‘50%")
        ]
        
        print(f"   éªŒè¯ç‚¹æ£€æŸ¥ç»“æœ:")
        all_passed = True
        for condition, description in verification_points:
            status = "âœ…" if condition else "âŒ"
            print(f"     {status} {description}")
            if not condition:
                all_passed = False
        
        # ä¿å­˜ç»“æœä¾›åç»­ä½¿ç”¨
        globals().update({
            'selection_test_results': selection_results,
            'difficulty_stats': difficulty_stats,
            'selection_accuracy': selection_accuracy
        })
        
        # æœ€ç»ˆéªŒè¯
        assert all_passed, "æ‰€æœ‰éªŒè¯ç‚¹éƒ½åº”è¯¥é€šè¿‡"
        
        print(f"\nâœ… éªŒè¯é€šè¿‡ï¼šGPT Agentèƒ½è‡ªåŠ¨é€‰æ‹©å¹¶ä½¿ç”¨å·¥å…·")
        print(f"\nğŸ¯ è‡ªåŠ¨å·¥å…·é€‰æ‹©éªŒè¯æ€»ç»“:")
        print(f"   âœ“ GPT Agentèƒ½å¤Ÿæ ¹æ®é—®é¢˜ç±»å‹è‡ªåŠ¨é€‰æ‹©å·¥å…·")
        print(f"   âœ“ GPT Agentèƒ½å¤Ÿå¤„ç†ä¸åŒéš¾åº¦çº§åˆ«çš„æ•°å­¦é—®é¢˜")
        print(f"   âœ“ GPT Agentèƒ½å¤Ÿæ‰§è¡Œå¤šæ­¥éª¤å¤æ‚è¿ç®—")
        print(f"   âœ“ å·¥å…·é€‰æ‹©å‡†ç¡®ç‡è¾¾åˆ°è¦æ±‚")
        print(f"   âœ“ éªŒè¯ç‚¹ï¼šGPT Agentèƒ½è‡ªåŠ¨é€‰æ‹©å¹¶ä½¿ç”¨å·¥å…· - é€šè¿‡")
        
    except Exception as e:
        print(f"\nâŒ éªŒè¯GPT Agentèƒ½è‡ªåŠ¨é€‰æ‹©å¹¶ä½¿ç”¨å·¥å…·å¤±è´¥: {e}")
```
