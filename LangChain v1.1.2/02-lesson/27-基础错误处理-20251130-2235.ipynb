## 2. 基础try-catch异常处理

```python
# 基础try-catch异常处理 - 独立代码块
from dotenv import load_dotenv
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import traceback

# 加载环境变量
load_dotenv()

print("🛡️ 基础try-catch异常处理:")
print("=" * 50)

# 1. 创建基础组件
print(f"🔧 1. 创建基础组件:")

api_key = os.getenv("OPENAI_API_KEY")
if api_key:
    print(f"   ✅ OpenAI API Key: 已配置")
else:
    print(f"   ❌ OpenAI API Key: 未配置")

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7,
    max_tokens=300,
    openai_api_key=api_key or "dummy_key"  # 允许测试错误情况
)

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "你是一个有用的AI助手。"),
    ("user", "问题: {query}")
])

output_parser = StrOutputParser()
chain = prompt_template | llm | output_parser

print(f"   基础组件创建完成")
print(f"   LLM模型: {llm.model_name}")
print(f"   链类型: {type(chain)}")

# 2. 无错误处理的基础调用
print(f"\n🧪 2. 无错误处理的基础调用:")

def basic_llm_call(query: str) -> str:
    """无错误处理的LLM调用（会崩溃）"""
    return chain.invoke({"query": query})

print(f"   基础调用函数定义完成")
print(f"   函数名: basic_llm_call")
print(f"   风险: 无异常处理，可能崩溃")

# 3. 带基础错误处理的调用
print(f"\n🛡️  3. 带基础错误处理的调用:")

def safe_llm_call(query: str) -> dict:
    """带基础错误处理的LLM调用"""
    try:
        # 尝试执行LLM调用
        result = chain.invoke({"query": query})
        
        return {
            "success": True,
            "response": result,
            "error": None
        }
        
    except Exception as e:
        # 捕获所有异常并返回错误信息
        return {
            "success": False,
            "response": None,
            "error": str(e),
            "error_type": type(e).__name__
        }

print(f"   安全调用函数定义完成")
print(f"   函数名: safe_llm_call")
print(f"   特性: 捕获异常，返回结构化结果")

# 4. 测试无错误处理的情况
print(f"\n🧪 4. 测试无错误处理的情况:")

if api_key:  # 只有在有有效API key时才测试
    try:
        print(f"   测试正常调用（无错误处理）...")
        basic_result = basic_llm_call("请简单介绍一下Python")
        print(f"   ✅ 基础调用成功: {len(basic_result)} 字符")
        basic_call_works = True
    except Exception as e:
        print(f"   ❌ 基础调用失败: {e}")
        basic_call_works = False
else:
    print(f"   ⚠️  跳过基础调用测试（无有效API Key）")
    basic_call_works = False

# 5. 测试有错误处理的情况
print(f"\n🛡️  5. 测试有错误处理的情况:")

try:
    print(f"   测试安全调用（有错误处理）...")
    safe_result = safe_llm_call("请简单介绍一下Python")
    
    print(f"   安全调用结果:")
    print(f"     成功: {safe_result['success']}")
    if safe_result['success']:
        print(f"     响应长度: {len(safe_result['response'])} 字符")
        print(f"     响应预览: {safe_result['response'][:80]}...")
    else:
        print(f"     错误类型: {safe_result.get('error_type', 'Unknown')}")
        print(f"     错误信息: {safe_result['error']}")
    
    safe_call_works = True
    
except Exception as e:
    print(f"   ❌ 安全调用测试失败: {e}")
    safe_call_works = False

# 6. 验证错误处理效果
print(f"\n🔍 6. 验证错误处理效果:")

def verify_error_handling() -> bool:
    """验证错误处理是否有效"""
    print(f"     验证错误处理效果...")
    
    # 检查安全调用是否不会崩溃
    if not safe_call_works:
        print(f"       ❌ 安全调用函数本身有问题")
        return False
    
    # 检查返回结构
    required_keys = ['success', 'response', 'error']
    if not all(key in safe_result for key in required_keys):
        print(f"       ❌ 返回结构不完整")
        return False
    
    # 检查错误情况下的行为
    if not safe_result['success']:
        if not safe_result['error']:
            print(f"       ❌ 失败情况下缺少错误信息")
            return False
        print(f"       ✅ 错误情况下正确返回错误信息")
    else:
        if not safe_result['response']:
            print(f"       ❌ 成功情况下缺少响应内容")
            return False
        print(f"       ✅ 成功情况下正确返回响应内容")
    
    print(f"       ✅ 错误处理验证通过")
    return True

# 执行验证
error_handling_works = verify_error_handling()

print(f"\n   验证结果:")
print(f"     基础调用: {'✅ 正常' if basic_call_works else '⚠️  跳过/失败'}")
print(f"     安全调用: {'✅ 正常' if safe_call_works else '❌ 异常'}")
print(f"     错误处理: {'✅ 有效' if error_handling_works else '❌ 无效'}")

# 7. 保存错误处理函数供后续使用
print(f"\n💾 7. 保存错误处理函数供后续使用:")

globals().update({
    'basic_llm_call': basic_llm_call,
    'safe_llm_call': safe_llm_call,
    'basic_call_works': basic_call_works,
    'safe_call_works': safe_call_works,
    'error_handling_works': error_handling_works,
    'safe_result': safe_result
})

print(f"   错误处理函数已保存到全局变量")
print(f"   safe_llm_call: 带错误处理的安全调用函数")

# 验证点：基础异常处理实现正确
assert safe_call_works, "安全调用函数应该正常工作"
assert error_handling_works, "错误处理应该有效"
assert 'success' in safe_result, "结果应该包含success字段"
assert 'error' in safe_result, "结果应该包含error字段"

print(f"\n✅ 验证通过：基础异常处理实现正确")
print(f"\n🎯 基础异常处理总结:")
print(f"   ✓ 创建无错误处理的基础调用函数")
print(f"   ✓ 实现带try-catch的安全调用函数")
print(f"   ✓ 验证错误处理不会导致应用崩溃")
print(f"   ✓ 确认错误情况下返回结构化信息")
print(f"   ✓ 准备测试具体错误场景")
```

## 3. 模拟常见错误场景

```python
# 模拟常见错误场景 - 独立代码块
from dotenv import load_dotenv
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import traceback

# 加载环境变量
load_dotenv()

print("🎭 模拟常见错误场景:")
print("=" * 50)

# 1. 重新创建基础组件（确保独立性）
print(f"🔧 1. 重新创建基础组件:")

api_key = os.getenv("OPENAI_API_KEY")
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "你是一个有用的AI助手。"),
    ("user", "问题: {query}")
])
output_parser = StrOutputParser()

print(f"   基础组件重新创建完成")

# 2. 定义错误处理测试函数
print(f"\n🛡️  2. 定义错误处理测试函数:")

def test_llm_with_error_handling(llm_instance, query: str, test_name: str) -> dict:
    """测试LLM调用的错误处理"""
    print(f"   测试: {test_name}")
    
    try:
        # 创建链
        chain = prompt_template | llm_instance | output_parser
        
        # 执行调用
        result = chain.invoke({"query": query})
        
        return {
            "test_name": test_name,
            "success": True,
            "response": result,
            "error": None,
            "error_type": None
        }
        
    except Exception as e:
        return {
            "test_name": test_name,
            "success": False,
            "response": None,
            "error": str(e),
            "error_type": type(e).__name__
        }

print(f"   错误处理测试函数定义完成")

# 3. 测试场景1: 无效API Key
print(f"\n🔑 3. 测试场景1: 无效API Key")

invalid_llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7,
    max_tokens=300,
    openai_api_key="sk-invalid-key-12345"  # 故意使用无效key
)

result1 = test_llm_with_error_handling(
    invalid_llm, 
    "请简单介绍一下Python", 
    "无效API Key测试"
)

print(f"   测试结果:")
print(f"     成功: {result1['success']}")
print(f"     错误类型: {result1.get('error_type', 'None')}")
if result1['error']:
    print(f"     错误信息: {result1['error'][:100]}...")

api_key_error_handled = not result1['success'] and result1['error'] is not None
print(f"     API Key错误处理: {'✅ 成功' if api_key_error_handled else '❌ 失败'}")

# 4. 测试场景2: 不存在的模型
print(f"\n🤖 4. 测试场景2: 不存在的模型")

if api_key:  # 只有在有有效API key时才测试
    invalid_model_llm = ChatOpenAI(
        model="gpt-invalid-model-123",  # 故意使用不存在的模型
        temperature=0.7,
        max_tokens=300,
        openai_api_key=api_key
    )
    
    result2 = test_llm_with_error_handling(
        invalid_model_llm, 
        "请简单介绍一下Python", 
        "无效模型测试"
    )
    
    print(f"   测试结果:")
    print(f"     成功: {result2['success']}")
    print(f"     错误类型: {result2.get('error_type', 'None')}")
    if result2['error']:
        print(f"     错误信息: {result2['error'][:100]}...")
    
    model_error_handled = not result2['success'] and result2['error'] is not None
    print(f"     模型错误处理: {'✅ 成功' if model_error_handled else '❌ 失败'}")
else:
    print(f"   ⚠️  跳过无效模型测试（无有效API Key）")
    result2 = {"success": False, "error": "Skipped"}
    model_error_handled = True  # 跳过不算失败

# 5. 测试场景3: 正常情况（验证错误处理不影响正常功能）
print(f"\n✅ 5. 测试场景3: 正常情况")

if api_key:
    normal_llm = ChatOpenAI(
        model="gpt-4o-mini",
        temperature=0.7,
        max_tokens=300,
        openai_api_key=api_key
    )
    
    result3 = test_llm_with_error_handling(
        normal_llm, 
        "请简单介绍一下Python", 
        "正常情况测试"
    )
    
    print(f"   测试结果:")
    print(f"     成功: {result3['success']}")
    if result3['success']:
        print(f"     响应长度: {len(result3['response'])} 字符")
        print(f"     响应预览: {result3['response'][:80]}...")
    
    normal_case_works = result3['success'] and result3['response'] is not None
    print(f"     正常情况处理: {'✅ 成功' if normal_case_works else '❌ 失败'}")
else:
    print(f"   ⚠️  跳过正常情况测试（无有效API Key）")
    result3 = {"success": False, "error": "Skipped"}
    normal_case_works = True  # 跳过不算失败

# 6. 测试场景4: 空输入处理
print(f"\n📝 6. 测试场景4: 空输入处理")

if api_key:
    normal_llm = ChatOpenAI(
        model="gpt-4o-mini",
        temperature=0.7,
        max_tokens=300,
        openai_api_key=api_key
    )
    
    result4 = test_llm_with_error_handling(
        normal_llm, 
        "",  # 空输入
        "空输入测试"
    )
    
    print(f"   测试结果:")
    print(f"     成功: {result4['success']}")
    print(f"     错误类型: {result4.get('error_type', 'None')}")
    if result4['error']:
        print(f"     错误信息: {result4['error'][:100]}...")
    
    empty_input_handled = True  # 空输入应该能处理（成功或失败都不崩溃）
    print(f"     空输入处理: {'✅ 成功' if empty_input_handled else '❌ 失败'}")
else:
    print(f"   ⚠️  跳过空输入测试（无有效API Key）")
    result4 = {"success": False, "error": "Skipped"}
    empty_input_handled = True

# 7. 综合错误处理验证
print(f"\n🔍 7. 综合错误处理验证:")

test_results = [result1, result2, result3, result4]
test_names = ["API Key错误", "模型错误", "正常情况", "空输入"]

print(f"   错误处理综合验证:")
handled_count = 0
total_count = len(test_results)

for i, (result, name) in enumerate(zip(test_results, test_names), 1):
    if result['success']:
        status = "✅ 成功执行"
        handled_count += 1
    elif result['error'] is not None:
        status = "✅ 错误被捕获"
        handled_count += 1
    else:
        status = "❌ 未处理异常"
    
    print(f"     {i}. {name}: {status}")

success_rate = handled_count / total_count
print(f"\n   处理成功率: {handled_count}/{total_count} ({success_rate:.1%})")

comprehensive_handling = success_rate >= 0.75
print(f"   综合处理: {'✅ 通过' if comprehensive_handling else '❌ 失败'}")

# 8. 保存测试结果供后续使用
print(f"\n💾 8. 保存测试结果供后续使用:")

globals().update({
    'test_llm_with_error_handling': test_llm_with_error_handling,
    'test_results': test_results,
    'api_key_error_handled': api_key_error_handled,
    'model_error_handled': model_error_handled,
    'normal_case_works': normal_case_works,
    'comprehensive_handling': comprehensive_handling
})

print(f"   错误场景测试结果已保存")
print(f"   test_results: 所有测试场景的结果")

# 验证点：常见错误场景处理正确
assert api_key_error_handled, "API Key错误应该被正确处理"
assert model_error_handled, "模型错误应该被正确处理"
assert normal_case_works or not api_key, "正常情况应该工作或无API Key"
assert comprehensive_handling, "综合错误处理应该有效"

print(f"\n✅ 验证通过：常见错误场景处理正确")
print(f"\n🎯 错误场景模拟总结:")
print(f"   ✓ 测试无效API Key错误处理")
print(f"   ✓ 测试不存在的模型错误处理")
print(f"   ✓ 验证正常情况不受错误处理影响")
print(f"   ✓ 测试空输入等边界情况")
print(f"   ✓ 综合验证错误处理覆盖率")
print(f"   ✓ 准备验证应用不会崩溃")
```

## 4. 验证应用不会崩溃

```python
# 验证应用不会崩溃 - 独立代码块
from dotenv import load_dotenv
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import traceback
import time

# 加载环境变量
load_dotenv()

print("🛡️ 验证应用不会崩溃:")
print("=" * 50)

# 1. 创建健壮的LLM调用函数
print(f"🔧 1. 创建健壮的LLM调用函数:")

def robust_llm_call(query: str, api_key_to_use: str = None, model_to_use: str = "gpt-4o-mini") -> dict:
    """健壮的LLM调用函数，处理各种错误情况"""
    try:
        # 参数验证
        if not query or not isinstance(query, str):
            return {
                "success": False,
                "response": None,
                "error": "Invalid query: query must be a non-empty string",
                "error_type": "ValueError"
            }
        
        # 创建LLM实例
        llm = ChatOpenAI(
            model=model_to_use,
            temperature=0.7,
            max_tokens=300,
            openai_api_key=api_key_to_use or os.getenv("OPENAI_API_KEY", "dummy_key")
        )
        
        # 创建链
        prompt_template = ChatPromptTemplate.from_messages([
            ("system", "你是一个有用的AI助手。"),
            ("user", "问题: {query}")
        ])
        output_parser = StrOutputParser()
        chain = prompt_template | llm | output_parser
        
        # 执行调用
        result = chain.invoke({"query": query})
        
        return {
            "success": True,
            "response": result,
            "error": None,
            "error_type": None
        }
        
    except Exception as e:
        return {
            "success": False,
            "response": None,
            "error": str(e),
            "error_type": type(e).__name__
        }

print(f"   健壮LLM调用函数定义完成")
print(f"   特性: 参数验证 + 异常捕获 + 结构化返回")

# 2. 崩溃测试用例定义
print(f"\n🧪 2. 崩溃测试用例定义:")

crash_test_cases = [
    {
        "name": "无效API Key",
        "query": "请简单介绍一下Python",
        "api_key": "sk-invalid-key-12345",
        "model": "gpt-4o-mini"
    },
    {
        "name": "无效模型",
        "query": "请简单介绍一下Python",
        "api_key": os.getenv("OPENAI_API_KEY", "dummy_key"),
        "model": "gpt-invalid-model-123"
    },
    {
        "name": "空查询",
        "query": "",
        "api_key": os.getenv("OPENAI_API_KEY", "dummy_key"),
        "model": "gpt-4o-mini"
    },
    {
        "name": "非字符串查询",
        "query": 123,
        "api_key": os.getenv("OPENAI_API_KEY", "dummy_key"),
        "model": "gpt-4o-mini"
    },
    {
        "name": "正常情况",
        "query": "请简单介绍一下Python",
        "api_key": os.getenv("OPENAI_API_KEY"),
        "model": "gpt-4o-mini"
    }
]

print(f"   崩溃测试用例数量: {len(crash_test_cases)}")
for i, case in enumerate(crash_test_cases, 1):
    print(f"     {i}. {case['name']}")

# 3. 执行崩溃测试
print(f"\n💥 3. 执行崩溃测试:")

crash_test_results = []
application_crashed = False

for i, test_case in enumerate(crash_test_cases, 1):
    print(f"\n   测试 {i}: {test_case['name']}")
    
    try:
        # 记录开始时间
        start_time = time.time()
        
        # 执行健壮调用
        result = robust_llm_call(
            query=test_case['query'],
            api_key_to_use=test_case['api_key'],
            model_to_use=test_case['model']
        )
        
        # 记录结束时间
        end_time = time.time()
        execution_time = end_time - start_time
        
        # 记录结果
        test_result = {
            "test_name": test_case['name'],
            "success": result['success'],
            "error": result['error'],
            "error_type": result['error_type'],
            "execution_time": execution_time,
            "crashed": False
        }
        
        crash_test_results.append(test_result)
        
        print(f"     ✅ 执行完成（未崩溃）")
        print(f"     成功: {result['success']}")
        print(f"     执行时间: {execution_time:.2f} 秒")
        if result['error']:
            print(f"     错误: {result['error_type']}: {result['error'][:50]}...")
        
    except Exception as e:
        # 如果连健壮函数都崩溃了，说明应用崩溃了
        print(f"     💥 应用崩溃: {e}")
        application_crashed = True
        
        test_result = {
            "test_name": test_case['name'],
            "success": False,
            "error": str(e),
            "error_type": "ApplicationCrash",
            "execution_time": 0,
            "crashed": True
        }
        
        crash_test_results.append(test_result)

# 4. 分析崩溃测试结果
print(f"\n📊 4. 分析崩溃测试结果:")

print(f"   崩溃测试结果分析:")
total_tests = len(crash_test_results)
crashed_tests = sum(1 for r in crash_test_results if r['crashed'])
successful_tests = sum(1 for r in crash_test_results if not r['crashed'])
handled_errors = sum(1 for r in crash_test_results if not r['crashed'] and not r['success'])
normal_success = sum(1 for r in crash_test_results if not r['crashed'] and r['success'])

print(f"     总测试数: {total_tests}")
print(f"     崩溃测试: {crashed_tests}")
print(f"     成功执行: {successful_tests}")
print(f"     错误处理: {handled_errors}")
print(f"     正常成功: {normal_success}")

# 计算稳定性指标
stability_rate = successful_tests / total_tests
error_handling_rate = handled_errors / total_tests if total_tests > 0 else 0

print(f"\n   稳定性指标:")
print(f"     应用稳定性: {stability_rate:.1%}")
print(f"     错误处理率: {error_handling_rate:.1%}")
print(f"     应用崩溃: {'❌ 是' if application_crashed else '✅ 否'}")

# 5. 验证应用健壮性
print(f"\n🔍 5. 验证应用健壮性:")

def verify_application_robustness() -> bool:
    """验证应用健壮性"""
    print(f"     验证应用健壮性...")
    
    # 检查是否有崩溃
    if application_crashed:
        print(f"       ❌ 应用发生崩溃")
        return False
    
    # 检查稳定性率
    if stability_rate < 0.8:
        print(f"       ❌ 应用稳定性不足: {stability_rate:.1%}")
        return False
    
    # 检查错误处理
    if error_handling_rate < 0.4:  # 至少40%的错误被处理
        print(f"       ❌ 错误处理不足: {error_handling_rate:.1%}")
        return False
    
    # 检查正常情况是否工作
    normal_test = next((r for r in crash_test_results if r['test_name'] == '正常情况'), None)
    if normal_test and not normal_test['crashed']:
        print(f"       ✅ 正常情况测试通过")
    elif not os.getenv("OPENAI_API_KEY"):
        print(f"       ⚠️  正常情况跳过（无API Key）")
    else:
        print(f"       ❌ 正常情况测试失败")
        return False
    
    print(f"       ✅ 应用健壮性验证通过")
    return True

# 执行验证
application_robust = verify_application_robustness()

# 6. 连续调用稳定性测试
print(f"\n🔄 6. 连续调用稳定性测试:")

try:
    print(f"   执行连续调用测试（10次）...")
    
    continuous_results = []
    for i in range(10):
        result = robust_llm_call(
            query=f"测试问题 {i+1}",
            api_key_to_use="sk-invalid-key",  # 使用无效key确保产生错误
            model_to_use="gpt-4o-mini"
        )
        continuous_results.append(result)
    
    # 检查连续调用稳定性
    all_handled = all(not r['success'] and r['error'] is not None for r in continuous_results)
    
    print(f"   连续调用结果:")
    print(f"     调用次数: {len(continuous_results)}")
    print(f"     全部处理: {'✅ 是' if all_handled else '❌ 否'}")
    print(f"     稳定性: {'✅ 优秀' if all_handled else '❌ 有问题'}")
    
    continuous_stability = all_handled
    
except Exception as e:
    print(f"   ❌ 连续调用测试失败: {e}")
    continuous_stability = False

# 7. 保存健壮性验证结果
print(f"\n💾 7. 保存健壮性验证结果:")

globals().update({
    'robust_llm_call': robust_llm_call,
    'crash_test_results': crash_test_results,
    'application_crashed': application_crashed,
    'stability_rate': stability_rate,
    'application_robust': application_robust,
    'continuous_stability': continuous_stability
})

print(f"   健壮性验证结果已保存")
print(f"   robust_llm_call: 健壮的LLM调用函数")

# 验证点：异常情况下GPT应用不会崩溃
assert not application_crashed, "应用不应该崩溃"
assert stability_rate >= 0.8, "应用稳定性应该足够高"
assert application_robust, "应用健壮性应该通过验证"
assert continuous_stability, "连续调用应该保持稳定"

print(f"\n✅ 验证通过：异常情况下GPT应用不会崩溃")
print(f"\n🎯 应用健壮性验证总结:")
print(f"   ✓ 创建健壮的LLM调用函数")
print(f"   ✓ 测试多种错误场景不崩溃")
print(f"   ✓ 验证错误捕获和返回机制")
print(f"   ✓ 确认应用稳定性达到要求")
print(f"   ✓ 测试连续调用稳定性")
print(f"   ✓ 准备完成学习总结")
```
