```python
# æ¶ˆæ¯ç®¡ç†æ“ä½œ - ç‹¬ç«‹ä»£ç å—
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_core.chat_history import ChatMessageHistory
from dotenv import load_dotenv
import os

# åŠ è½½ç¯å¢ƒå˜é‡
load_dotenv()

print("ğŸ”§ æ¶ˆæ¯ç®¡ç†æ“ä½œæµ‹è¯•:")
print("=" * 50)

# åˆ›å»ºæ–°çš„æ¶ˆæ¯å†å²
history = ChatMessageHistory()

# æ·»åŠ åˆå§‹æ¶ˆæ¯
history.add_message(SystemMessage(content="ä½ æ˜¯ä¸€ä¸ªä¸“ä¸šçš„Pythonç¼–ç¨‹åŠ©æ‰‹ã€‚"))
history.add_message(HumanMessage(content="å¦‚ä½•å®šä¹‰ä¸€ä¸ªPythonå‡½æ•°ï¼Ÿ"))
history.add_message(AIMessage(content="ä½¿ç”¨defå…³é”®å­—å®šä¹‰å‡½æ•°ï¼Œä¾‹å¦‚ï¼šdef my_function(): pass"))

print(f"ğŸ“ åˆå§‹æ¶ˆæ¯å†å²:")
for i, message in enumerate(history.messages, 1):
    print(f"   {i}. [{type(message).__name__}] {message.content}")

# 1. æ·»åŠ ç”¨æˆ·æ¶ˆæ¯
print(f"\nâ• 1. æ·»åŠ ç”¨æˆ·æ¶ˆæ¯:")

new_user_question = HumanMessage(content="èƒ½ç»™æˆ‘ä¸€ä¸ªæ›´è¯¦ç»†çš„ä¾‹å­å—ï¼Ÿ")
history.add_message(new_user_question)

print(f"   æ·»åŠ çš„æ–°é—®é¢˜: {new_user_question.content}")
print(f"   å½“å‰æ¶ˆæ¯æ€»æ•°: {len(history.messages)}")

# 2. æ·»åŠ AIå›å¤
print(f"\nğŸ¤– 2. æ·»åŠ AIå›å¤:")

ai_response = AIMessage(content="""å½“ç„¶ï¼è¿™é‡Œæ˜¯ä¸€ä¸ªè¯¦ç»†çš„Pythonå‡½æ•°ä¾‹å­ï¼š

def calculate_area(length, width):
    """
    è®¡ç®—çŸ©å½¢é¢ç§¯
    
    Args:
        length (float): é•¿åº¦
        width (float): å®½åº¦
    
    Returns:
        float: é¢ç§¯
    """
    area = length * width
    return area

# ä½¿ç”¨ç¤ºä¾‹
result = calculate_area(5, 3)
print(f"é¢ç§¯æ˜¯: {result}")""")

history.add_message(ai_response)

print(f"   æ·»åŠ çš„AIå›å¤é•¿åº¦: {len(ai_response.content)} å­—ç¬¦")
print(f"   å½“å‰æ¶ˆæ¯æ€»æ•°: {len(history.messages)}")

# 3. æ¶ˆæ¯æŸ¥è¯¢å’Œè·å–
print(f"\nğŸ” 3. æ¶ˆæ¯æŸ¥è¯¢å’Œè·å–:")

# è·å–æ‰€æœ‰æ¶ˆæ¯
all_messages = history.messages
print(f"   æ‰€æœ‰æ¶ˆæ¯æ•°é‡: {len(all_messages)}")

# è·å–æœ€åä¸€æ¡æ¶ˆæ¯
last_message = history.messages[-1]
print(f"   æœ€åä¸€æ¡æ¶ˆæ¯ç±»å‹: {type(last_message).__name__}")
print(f"   æœ€åä¸€æ¡æ¶ˆæ¯å†…å®¹: {last_message.content[:50]}...")

# è·å–äººç±»æ¶ˆæ¯
human_messages = [msg for msg in history.messages if isinstance(msg, HumanMessage)]
print(f"   äººç±»æ¶ˆæ¯æ•°é‡: {len(human_messages)}")
for i, msg in enumerate(human_messages, 1):
    print(f"     äººç±»æ¶ˆæ¯{i}: {msg.content}")

# è·å–AIæ¶ˆæ¯
ai_messages = [msg for msg in history.messages if isinstance(msg, AIMessage)]
print(f"   AIæ¶ˆæ¯æ•°é‡: {len(ai_messages)}")
for i, msg in enumerate(ai_messages, 1):
        print(f"     AIæ¶ˆæ¯{i}: {msg.content[:50]}...")

# 4. æ¶ˆæ¯ç»Ÿè®¡å’Œåˆ†æ
print(f"\nğŸ“Š 4. æ¶ˆæ¯ç»Ÿè®¡å’Œåˆ†æ:")

# æŒ‰ç±»å‹ç»Ÿè®¡
from collections import Counter
message_types = [type(msg).__name__ for msg in history.messages]
type_counts = Counter(message_types)

print(f"   æ¶ˆæ¯ç±»å‹åˆ†å¸ƒ:")
for msg_type, count in type_counts.items():
    print(f"     {msg_type}: {count} æ¡")

# å†…å®¹é•¿åº¦ç»Ÿè®¡
content_lengths = [len(msg.content) for msg in history.messages]
print(f"\n   å†…å®¹é•¿åº¦ç»Ÿè®¡:")
print(f"     æ€»å­—ç¬¦æ•°: {sum(content_lengths)}")
print(f"     å¹³å‡é•¿åº¦: {sum(content_lengths)/len(content_lengths):.1f}")
print(f"     æœ€é•¿æ¶ˆæ¯: {max(content_lengths)} å­—ç¬¦")
print(f"     æœ€çŸ­æ¶ˆæ¯: {min(content_lengths)} å­—ç¬¦")

# 5. æ¶ˆæ¯æ ¼å¼åŒ–
print(f"\nğŸ“‹ 5. æ¶ˆæ¯æ ¼å¼åŒ–:")

def format_chat_history(history):
    """æ ¼å¼åŒ–èŠå¤©å†å²ä¸ºå¯è¯»æ ¼å¼"""
    formatted = ""
    for i, message in enumerate(history.messages, 1):
        msg_type = type(message).__name__
        if msg_type == "SystemMessage":
            formatted += f"[ç³»ç»Ÿ] {message.content}\n"
        elif msg_type == "HumanMessage":
            formatted += f"[ç”¨æˆ·] {message.content}\n"
        elif msg_type == "AIMessage":
            formatted += f"[AI] {message.content}\n"
    return formatted

formatted_history = format_chat_history(history)
print(f"   æ ¼å¼åŒ–å¯¹è¯å†å²:")
print(f"   {formatted_history}")

# éªŒè¯ç‚¹ï¼šæ¶ˆæ¯ç®¡ç†æ“ä½œæ­£ç¡®
assert len(history.messages) == 5, "æ¶ˆæ¯æ€»æ•°ä¸æ­£ç¡®"
assert len(human_messages) == 2, "äººç±»æ¶ˆæ¯æ•°é‡ä¸æ­£ç¡®"
assert len(ai_messages) == 2, "AIæ¶ˆæ¯æ•°é‡ä¸æ­£ç¡®"
assert sum(content_lengths) > 0, "æ€»å­—ç¬¦æ•°åº”è¯¥å¤§äº0"
assert len(formatted_history) > 0, "æ ¼å¼åŒ–å†å²ä¸åº”ä¸ºç©º"

print(f"\nâœ… éªŒè¯é€šè¿‡ï¼šæ¶ˆæ¯ç®¡ç†æ“ä½œæ­£ç¡®")
```

## 3. ä¸é€šä¹‰åƒé—®é›†æˆä½¿ç”¨

```python
# ä¸é€šä¹‰åƒé—®é›†æˆä½¿ç”¨ - ç‹¬ç«‹ä»£ç å—
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_core.chat_history import ChatMessageHistory
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from dotenv import load_dotenv
import os

# åŠ è½½ç¯å¢ƒå˜é‡
load_dotenv()

print("ğŸ¤– ä¸é€šä¹‰åƒé—®é›†æˆä½¿ç”¨æµ‹è¯•:")
print("=" * 50)

openai_api_key = os.getenv('OPENAI_API_KEY')
if not openai_api_key:
    print("âŒ OpenAI API Key æœªé…ç½®")
else:
    try:
        # åˆ›å»º LLM
        llm = ChatOpenAI(
            model="gpt-4o-mini",
            temperature=0.7,
            max_tokens=100
        )
        
        print(f"ğŸ“ åˆ›å»ºçš„ç»„ä»¶:")
        print(f"   1. LLM: {type(llm).__name__}")
        print(f"   2. æ¨¡å‹: {llm.model_name}")
        
        # åˆ›å»ºæ¶ˆæ¯å†å²
        history = ChatMessageHistory()
        
        # æ·»åŠ ç³»ç»Ÿæ¶ˆæ¯
        history.add_message(SystemMessage(content="ä½ æ˜¯ä¸€ä¸ªä¸“ä¸šçš„LangChainæŠ€æœ¯é¡¾é—®ï¼Œä¸“é—¨å›ç­”å…³äºLangChainæ¡†æ¶çš„é—®é¢˜ã€‚"))
        
        print(f"\nğŸ—ï¸  åˆ›å»ºå¯¹è¯å†å²:")
        print(f"   ç³»ç»Ÿæ¶ˆæ¯å·²è®¾ç½®: {history.messages[0].content}")
        
        # åˆ›å»ºå¸¦å†å²æ¶ˆæ¯çš„æç¤ºæ¨¡æ¿
        prompt = ChatPromptTemplate.from_messages([
            ("system", "ä½ æ˜¯ä¸€ä¸ªä¸“ä¸šçš„LangChainæŠ€æœ¯é¡¾é—®ã€‚"),
            MessagesPlaceholder(variable_name="history"),
            ("human", "{input}")
        ])
        
        print(f"\nğŸ”— åˆ›å»ºæç¤ºæ¨¡æ¿:")
        print(f"   æ¨¡æ¿ç±»å‹: {type(prompt)}")
        print(f"   åŒ…å«å†å²æ¶ˆæ¯å ä½ç¬¦: {MessagesPlaceholder in [type(msg) for msg in prompt.messages if hasattr(msg, 'variable')]}")
        
        # åˆ›å»ºå¤„ç†é“¾
        chain = prompt | llm
        
        print(f"\nğŸ”— åˆ›å»ºå¤„ç†é“¾:")
        print(f"   é“¾ç±»å‹: {type(chain)}")
        
        # 1. ç¬¬ä¸€æ¬¡å¯¹è¯
        print(f"\nğŸ’¬ 1. ç¬¬ä¸€æ¬¡å¯¹è¯:")
        
        first_question = "ä»€ä¹ˆæ˜¯LangChainï¼Ÿ"
        print(f"   ç”¨æˆ·é—®é¢˜: {first_question}")
        
        # æ·»åŠ ç”¨æˆ·æ¶ˆæ¯åˆ°å†å²
        history.add_message(HumanMessage(content=first_question))
        
        # è°ƒç”¨é“¾è·å–å›å¤
        response = chain.invoke({
            "history": history.messages,
            "input": first_question
        })
        
        # æ·»åŠ AIå›å¤åˆ°å†å²
        history.add_message(response)
        
        print(f"   AIå›å¤: {response.content}")
        print(f"   å†å²æ¶ˆæ¯æ•°: {len(history.messages)}")
        
        # 2. ç¬¬äºŒæ¬¡å¯¹è¯ï¼ˆå¸¦ä¸Šä¸‹æ–‡ï¼‰
        print(f"\nğŸ’¬ 2. ç¬¬äºŒæ¬¡å¯¹è¯ï¼ˆå¸¦ä¸Šä¸‹æ–‡ï¼‰:")
        
        second_question = "å®ƒæœ‰å“ªäº›ä¸»è¦ç»„ä»¶ï¼Ÿ"
        print(f"   ç”¨æˆ·é—®é¢˜: {second_question}")
        print(f"   å½“å‰ä¸Šä¸‹æ–‡æ¶ˆæ¯æ•°: {len(history.messages)}")
        
        # æ·»åŠ ç”¨æˆ·æ¶ˆæ¯åˆ°å†å²
        history.add_message(HumanMessage(content=second_question))
        
        # è°ƒç”¨é“¾è·å–å›å¤ï¼ˆåŒ…å«å†å²ä¸Šä¸‹æ–‡ï¼‰
        response2 = chain.invoke({
            "history": history.messages,
            "input": second_question
        })
        
        # æ·»åŠ AIå›å¤åˆ°å†å²
        history.add_message(response2)
        
        print(f"   AIå›å¤: {response2.content}")
        print(f"   å†å²æ¶ˆæ¯æ•°: {len(history.messages)}")
        
        # 3. ç¬¬ä¸‰æ¬¡å¯¹è¯ï¼ˆæµ‹è¯•ä¸Šä¸‹æ–‡ç†è§£ï¼‰
        print(f"\nğŸ’¬ 3. ç¬¬ä¸‰æ¬¡å¯¹è¯ï¼ˆæµ‹è¯•ä¸Šä¸‹æ–‡ç†è§£ï¼‰:")
        
        third_question = "èƒ½è¯¦ç»†ä»‹ç»ä¸€ä¸‹LCELå—ï¼Ÿ"
        print(f"   ç”¨æˆ·é—®é¢˜: {third_question}")
        
        # æ·»åŠ ç”¨æˆ·æ¶ˆæ¯åˆ°å†å²
        history.add_message(HumanMessage(content=third_question))
        
        # è°ƒç”¨é“¾è·å–å›å¤
        response3 = chain.invoke({
            "history": history.messages,
            "input": third_question
        })
        
        # æ·»åŠ AIå›å¤åˆ°å†å²
        history.add_message(response3)
        
        print(f"   AIå›å¤: {response3.content}")
        print(f"   å†å²æ¶ˆæ¯æ•°: {len(history.messages)}")
        
        # 4. æ˜¾ç¤ºå®Œæ•´å¯¹è¯å†å²
        print(f"\nğŸ“‹ 4. å®Œæ•´å¯¹è¯å†å²:")
        
        for i, message in enumerate(history.messages, 1):
            msg_type = type(message).__name__
            content_preview = message.content[:50] + "..." if len(message.content) > 50 else message.content
            print(f"   {i}. [{msg_type}] {content_preview}")
        
        # 5. æµ‹è¯•ä¸Šä¸‹æ–‡ç†è§£
        print(f"\nğŸ§ª 5. æµ‹è¯•ä¸Šä¸‹æ–‡ç†è§£:")
        
        context_question = "åˆšæ‰æˆ‘ä»¬è®¨è®ºçš„LCELæ˜¯ä»€ä¹ˆçš„ç¼©å†™ï¼Ÿ"
        print(f"   ä¸Šä¸‹æ–‡æµ‹è¯•é—®é¢˜: {context_question}")
        
        history.add_message(HumanMessage(content=context_question))
        
        context_response = chain.invoke({
            "history": history.messages,
            "input": context_question
        })
        
        print(f"   ä¸Šä¸‹æ–‡å›å¤: {context_response.content}")
        
        # éªŒè¯ç‚¹ï¼šä¸é€šä¹‰åƒé—®é›†æˆæ­£ç¡®
        assert len(history.messages) >= 7, "å¯¹è¯å†å²æ¶ˆæ¯æ•°é‡ä¸è¶³"
        assert isinstance(response, AIMessage), "ç¬¬ä¸€æ¬¡å›å¤ä¸æ˜¯AIæ¶ˆæ¯ç±»å‹"
        assert isinstance(response2, AIMessage), "ç¬¬äºŒæ¬¡å›å¤ä¸æ˜¯AIæ¶ˆæ¯ç±»å‹"
        assert isinstance(response3, AIMessage), "ç¬¬ä¸‰æ¬¡å›å¤ä¸æ˜¯AIæ¶ˆæ¯ç±»å‹"
        assert len(response.content) > 0, "ç¬¬ä¸€æ¬¡å›å¤å†…å®¹ä¸ºç©º"
        assert len(response2.content) > 0, "ç¬¬äºŒæ¬¡å›å¤å†…å®¹ä¸ºç©º"
        assert len(response3.content) > 0, "ç¬¬ä¸‰æ¬¡å›å¤å†…å®¹ä¸ºç©º"
        
        print(f"\nâœ… éªŒè¯é€šè¿‡ï¼šä¸é€šä¹‰åƒé—®é›†æˆä½¿ç”¨æ­£ç¡®")
        
    except Exception as e:
        print(f"\nâŒ é€šä¹‰åƒé—®é›†æˆå¤±è´¥: {e}")
```

## 4. å†å²æ¶ˆæ¯ç®¡ç†ç­–ç•¥

```python
# å†å²æ¶ˆæ¯ç®¡ç†ç­–ç•¥ - ç‹¬ç«‹ä»£ç å—
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_core.chat_history import ChatMessageHistory
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from dotenv import load_dotenv
import os
import time

# åŠ è½½ç¯å¢ƒå˜é‡
load_dotenv()

print("ğŸ“Š å†å²æ¶ˆæ¯ç®¡ç†ç­–ç•¥æµ‹è¯•:")
print("=" * 50)

openai_api_key = os.getenv('OPENAI_API_KEY')
if not openai_api_key:
    print("âŒ OpenAI API Key æœªé…ç½®")
else:
    try:
        # åˆ›å»ºåŸºç¡€ç»„ä»¶
        llm = ChatOpenAI(
            model="gpt-4o-mini",
            temperature=0.7,
            max_tokens=60
        )
        
        prompt = ChatPromptTemplate.from_messages([
            MessagesPlaceholder(variable_name="history"),
            ("human", "{input}")
        ])
        
        chain = prompt | llm
        
        print(f"ğŸ“ åˆ›å»ºçš„ç®¡ç†ç»„ä»¶:")
        print(f"   1. LLM: {type(llm).__name__}")
        print(f"   2. æç¤ºæ¨¡æ¿: {type(prompt)}")
        print(f"   3. å¤„ç†é“¾: {type(chain)}")
        
        # 1. æ¶ˆæ¯æ•°é‡é™åˆ¶ç­–ç•¥
        print(f"\nğŸ§ª 1. æ¶ˆæ¯æ•°é‡é™åˆ¶ç­–ç•¥:")
        
        def limit_message_count(history, max_messages=10):
            """é™åˆ¶æ¶ˆæ¯æ•°é‡çš„ç­–ç•¥"""
            messages = history.messages
            
            if len(messages) <= max_messages:
                return messages
            
            # ä¿ç•™ç³»ç»Ÿæ¶ˆæ¯å’Œæœ€è¿‘çš„max_messagesæ¡æ¶ˆæ¯
            system_messages = [msg for msg in messages if isinstance(msg, SystemMessage)]
            other_messages = [msg for msg in messages if not isinstance(msg, SystemMessage)]
            
            # ä¿ç•™æœ€è¿‘çš„max_messages - len(system_messages)æ¡å…¶ä»–æ¶ˆæ¯
            recent_others = other_messages[-(max_messages - len(system_messages)):]
            
            return system_messages + recent_others
        
        # æµ‹è¯•æ•°é‡é™åˆ¶
        history = ChatMessageHistory()
        history.add_message(SystemMessage(content="ä½ æ˜¯ä¸€ä¸ªAIåŠ©æ‰‹ã€‚"))
        
        # æ·»åŠ 20æ¡æ¶ˆæ¯
        for i in range(10):
            history.add_message(HumanMessage(content=f"é—®é¢˜{i+1}"))
            history.add_message(AIMessage(content=f"å›ç­”{i+1}"))
        
        print(f"   åŸå§‹æ¶ˆæ¯æ•°: {len(history.messages)}")
        
        limited_messages = limit_message_count(history, max_messages=8)
        print(f"   é™åˆ¶åæ¶ˆæ¯æ•°: {len(limited_messages)}")
        print(f"   ä¿ç•™çš„ç³»ç»Ÿæ¶ˆæ¯: {sum(1 for msg in limited_messages if isinstance(msg, SystemMessage))}")
        print(f"   ä¿ç•™çš„å…¶ä»–æ¶ˆæ¯: {sum(1 for msg in limited_messages if not isinstance(msg, SystemMessage))}")
        
        # 2. æ¶ˆæ¯é•¿åº¦é™åˆ¶ç­–ç•¥
        print(f"\nğŸ§ª 2. æ¶ˆæ¯é•¿åº¦é™åˆ¶ç­–ç•¥:")
        
        def limit_message_length(history, max_total_chars=500):
            """é™åˆ¶æ¶ˆæ¯æ€»é•¿åº¦çš„ç­–ç•¥"""
            messages = history.messages
            
            total_chars = sum(len(msg.content) for msg in messages)
            if total_chars <= max_total_chars:
                return messages
            
            # ä»æœ€æ–°æ¶ˆæ¯å¼€å§‹ï¼Œä¿ç•™ç³»ç»Ÿæ¶ˆæ¯
            system_messages = [msg for msg in messages if isinstance(msg, SystemMessage)]
            other_messages = [msg for msg in messages if not isinstance(msg, SystemMessage)]
            
            result = system_messages.copy()
            current_chars = sum(len(msg.content) for msg in result)
            
            # ä»æœ€æ–°æ¶ˆæ¯å¼€å§‹æ·»åŠ 
            for msg in reversed(other_messages):
                if current_chars + len(msg.content) <= max_total_chars:
                    result.insert(len(system_messages), msg)  # æ’å…¥åˆ°ç³»ç»Ÿæ¶ˆæ¯å
                    current_chars += len(msg.content)
                else:
                    break
            
            return result
        
        # æµ‹è¯•é•¿åº¦é™åˆ¶
        length_limited = limit_message_length(history, max_total_chars=200)
        
        length_total = sum(len(msg.content) for msg in length_limited)
        print(f"   é•¿åº¦é™åˆ¶åæ¶ˆæ¯æ•°: {len(length_limited)}")
        print(f"   æ€»å­—ç¬¦æ•°: {length_total}")
        print(f"   æ˜¯å¦æ»¡è¶³é™åˆ¶: {length_total <= 200}")
        
        # 3. æ—¶é—´çª—å£ç­–ç•¥
        print(f"\nğŸ§ª 3. æ—¶é—´çª—å£ç­–ç•¥:")
        
        def time_window_filter(history, max_age_minutes=5):
            """åŸºäºæ—¶é—´çª—å£çš„æ¶ˆæ¯è¿‡æ»¤ç­–ç•¥"""
            current_time = time.time()
            max_age_seconds = max_age_minutes * 60
            
            messages = history.messages
            
            # ä¿ç•™æ‰€æœ‰ç³»ç»Ÿæ¶ˆæ¯
            system_messages = [msg for msg in messages if isinstance(msg, SystemMessage)]
            
            # æ¨¡æ‹Ÿæ—¶é—´æˆ³ï¼ˆå®é™…åº”ç”¨ä¸­æ¶ˆæ¯åº”è¯¥æœ‰æ—¶é—´æˆ³ï¼‰
            # è¿™é‡Œæˆ‘ä»¬å‡è®¾æ¶ˆæ¯æ˜¯æŒ‰æ—¶é—´é¡ºåºæ·»åŠ çš„
            other_messages = [msg for msg in messages if not isinstance(msg, SystemMessage)]
            
            # ç®€åŒ–ç­–ç•¥ï¼šä¿ç•™æœ€è¿‘çš„æ¶ˆæ¯ï¼ˆæ¨¡æ‹Ÿæ—¶é—´çª—å£ï¼‰
            recent_count = min(10, len(other_messages))  # ä¿ç•™æœ€è¿‘10æ¡
            recent_others = other_messages[-recent_count:] if recent_count > 0 else []
            
            return system_messages + recent_others
        
        # æµ‹è¯•æ—¶é—´çª—å£
        time_filtered = time_window_filter(history, max_age_minutes=5)
        print(f"   æ—¶é—´çª—å£è¿‡æ»¤åæ¶ˆæ¯æ•°: {len(time_filtered)}")
        print(f"   ä¿ç•™ç³»ç»Ÿæ¶ˆæ¯: {sum(1 for msg in time_filtered if isinstance(msg, SystemMessage))}")
        print(f"   ä¿ç•™å…¶ä»–æ¶ˆæ¯: {sum(1 for msg in time_filtered if not isinstance(msg, SystemMessage))}")
        
        # 4. æ™ºèƒ½æ‘˜è¦ç­–ç•¥
        print(f"\nğŸ§ª 4. æ™ºèƒ½æ‘˜è¦ç­–ç•¥:")
        
        def summarize_old_messages(history, summary_threshold=15):
            """æ™ºèƒ½æ‘˜è¦æ—§æ¶ˆæ¯çš„ç­–ç•¥"""
            messages = history.messages
            
            if len(messages) <= summary_threshold:
                return messages
            
            # åˆ†ç¦»ç³»ç»Ÿæ¶ˆæ¯å’Œå…¶ä»–æ¶ˆæ¯
            system_messages = [msg for msg in messages if isinstance(msg, SystemMessage)]
            other_messages = [msg for msg in messages if not isinstance(msg, SystemMessage)]
            
            # ä¿ç•™ç³»ç»Ÿæ¶ˆæ¯
            result = system_messages.copy()
            
            # ä¿ç•™æœ€æ–°çš„8æ¡æ¶ˆæ¯
            recent_messages = other_messages[-8:] if len(other_messages) > 8 else other_messages
            result.extend(recent_messages)
            
            # å¦‚æœè¿˜æœ‰æ—§æ¶ˆæ¯ï¼Œåˆ›å»ºæ‘˜è¦
            old_messages = other_messages[:-8] if len(other_messages) > 8 else []
            if old_messages:
                # åˆ›å»ºç®€å•çš„æ–‡æœ¬æ‘˜è¦
                summary_text = f"ä¹‹å‰æœ‰{len(old_messages)//2}è½®å¯¹è¯å·²è¢«æ‘˜è¦ã€‚"
                summary_msg = SystemMessage(content=summary_text)
                result.insert(len(system_messages), summary_msg)
            
            return result
        
        # æµ‹è¯•æ™ºèƒ½æ‘˜è¦
        summarized = summarize_old_messages(history, summary_threshold=10)
        print(f"   æ™ºèƒ½æ‘˜è¦åæ¶ˆæ¯æ•°: {len(summarized)}")
        
        summary_messages = [msg for msg in summarized if "æ‘˜è¦" in msg.content]
        print(f"   åŒ…å«æ‘˜è¦æ¶ˆæ¯: {len(summary_messages)}")
        
        # 5. ç­–ç•¥å¯¹æ¯”
        print(f"\nğŸ“Š 5. ç­–ç•¥å¯¹æ¯”:")
        print(f"   åŸå§‹æ¶ˆæ¯æ•°: {len(history.messages)}")
        print(f"   æ•°é‡é™åˆ¶ç­–ç•¥: {len(limited_messages)} æ¡æ¶ˆæ¯")
        print(f"   é•¿åº¦é™åˆ¶ç­–ç•¥: {len(length_limited)} æ¡æ¶ˆæ¯")
        print(f"   æ—¶é—´çª—å£ç­–ç•¥: {len(time_filtered)} æ¡æ¶ˆæ¯")
        print(f"   æ™ºèƒ½æ‘˜è¦ç­–ç•¥: {len(summarized)} æ¡æ¶ˆæ¯")
        
        # éªŒè¯ç‚¹ï¼šå†å²æ¶ˆæ¯ç®¡ç†ç­–ç•¥æ­£ç¡®
        assert len(limited_messages) <= 8, "æ•°é‡é™åˆ¶ç­–ç•¥å¤±è´¥"
        assert sum(len(msg.content) for msg in length_limited) <= 200, "é•¿åº¦é™åˆ¶ç­–ç•¥å¤±è´¥"
        assert len(time_filtered) < len(history.messages), "æ—¶é—´çª—å£ç­–ç•¥åº”è¯¥å‡å°‘æ¶ˆæ¯æ•°"
        assert len(summarized) < len(history.messages), "æ™ºèƒ½æ‘˜è¦ç­–ç•¥åº”è¯¥å‡å°‘æ¶ˆæ¯æ•°"
        
        print(f"\nâœ… éªŒè¯é€šè¿‡ï¼šå†å²æ¶ˆæ¯ç®¡ç†ç­–ç•¥æ­£ç¡®")
        
    except Exception as e:
        print(f"\nâŒ å†å²æ¶ˆæ¯ç®¡ç†ç­–ç•¥æµ‹è¯•å¤±è´¥: {e}")
```
