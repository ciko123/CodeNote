```python
# éªŒè¯å·¥å…·èƒ½æ­£ç¡®æ‰§è¡Œå¹¶è¿”å›GPTç»“æœ - ç‹¬ç«‹ä»£ç å—
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os
import json

# åŠ è½½ç¯å¢ƒå˜é‡
load_dotenv()

print("ğŸ¤– éªŒè¯å·¥å…·èƒ½æ­£ç¡®æ‰§è¡Œå¹¶è¿”å›GPTç»“æœ:")
print("=" * 50)

openai_api_key = os.getenv('OPENAI_API_KEY')
if not openai_api_key:
    print("âŒ OpenAI API Key æœªé…ç½®")
else:
    try:
        # 1. å‡†å¤‡å·¥å…·é›†åˆ
        print(f"ğŸ”§ 1. å‡†å¤‡å·¥å…·é›†åˆ:")
        
        # æ£€æŸ¥æ˜¯å¦æœ‰å·²éªŒè¯çš„å·¥å…·
        if 'verified_tools' in globals():
            calculator_tools = globals()['verified_tools']
            print(f"   ä½¿ç”¨å·²éªŒè¯çš„å·¥å…·é›†åˆ: {len(calculator_tools)} ä¸ªå·¥å…·")
        else:
            print(f"   é‡æ–°åˆ›å»ºå·¥å…·é›†åˆ...")
            
            @tool
            def åŠ æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å’Œï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
                result = a + b
                return f"{a} + {b} = {result}"
            
            @tool
            def å‡æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å·®ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
                result = a - b
                return f"{a} - {b} = {result}"
            
            @tool
            def ä¹˜æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„ç§¯ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—"""
                result = a * b
                return f"{a} Ã— {b} = {result}"
            
            @tool
            def é™¤æ³•(a: float, b: float) -> str:
                """è®¡ç®—ä¸¤ä¸ªæ•°å­—çš„å•†ï¼Œæ”¯æŒå°æ•°ç‚¹è¿ç®—ã€‚é™¤æ•°ä¸èƒ½ä¸ºé›¶"""
                if b == 0:
                    return "é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸ºé›¶"
                result = a / b
                return f"{a} Ã· {b} = {result}"
            
            calculator_tools = {
                'addition': åŠ æ³•,
                'subtraction': å‡æ³•,
                'multiplication': ä¹˜æ³•,
                'division': é™¤æ³•
            }
        
        tools_list = list(calculator_tools.values())
        print(f"   å·¥å…·åˆ—è¡¨å‡†å¤‡å®Œæˆ: {len(tools_list)} ä¸ªå·¥å…·")
        
        # 2. åˆ›å»ºGPT LLM
        print(f"\nğŸ¤– 2. åˆ›å»ºGPT LLM:")
        
        llm = ChatOpenAI(
            model="gpt-4o-mini",
            temperature=0.1,  # ä½æ¸©åº¦ç¡®ä¿å‡†ç¡®è®¡ç®—
            max_tokens=100
        )
        
        print(f"   LLMé…ç½®:")
        print(f"     æ¨¡å‹: {llm.model_name}")
        print(f"     æ¸©åº¦: {llm.temperature}")
        print(f"     æœ€å¤§ä»¤ç‰Œ: {llm.max_tokens}")
        
        # 3. å·¥å…·ç»‘å®šæµ‹è¯•
        print(f"\nğŸ”— 3. å·¥å…·ç»‘å®šæµ‹è¯•:")
        
        try:
            # å°†å·¥å…·ç»‘å®šåˆ°LLM
            llm_with_tools = llm.bind_tools(tools_list)
            print(f"   å·¥å…·ç»‘å®šæˆåŠŸ: {type(llm_with_tools)}")
            print(f"   ç»‘å®šçš„å·¥å…·æ•°é‡: {len(tools_list)}")
        except Exception as e:
            print(f"   å·¥å…·ç»‘å®šå¤±è´¥: {e}")
            llm_with_tools = llm
        
        # 4. GPTå·¥å…·è°ƒç”¨æµ‹è¯•
        print(f"\nğŸ§ª 4. GPTå·¥å…·è°ƒç”¨æµ‹è¯•:")
        
        # åˆ›å»ºæµ‹è¯•æç¤º
        test_prompts = [
            "è¯·è®¡ç®— 25 + 17 çš„ç»“æœ",
            "è¯·è®¡ç®— 100 - 37 çš„ç»“æœ",
            "è¯·è®¡ç®— 8 Ã— 9 çš„ç»“æœ",
            "è¯·è®¡ç®— 144 Ã· 12 çš„ç»“æœ"
        ]
        
        print(f"   æ‰§è¡ŒGPTå·¥å…·è°ƒç”¨æµ‹è¯•:")
        gpt_test_results = []
        
        for i, prompt in enumerate(test_prompts, 1):
            print(f"\n     æµ‹è¯• {i}: {prompt}")
            
            try:
                # è°ƒç”¨GPT
                response = llm_with_tools.invoke(prompt)
                
                print(f"       GPTå“åº”ç±»å‹: {type(response)}")
                print(f"       GPTå“åº”å†…å®¹: {response.content[:100]}...")
                
                # æ£€æŸ¥æ˜¯å¦æœ‰å·¥å…·è°ƒç”¨
                if hasattr(response, 'tool_calls') and response.tool_calls:
                    print(f"       æ£€æµ‹åˆ°å·¥å…·è°ƒç”¨: {len(response.tool_calls)} ä¸ª")
                    for j, tool_call in enumerate(response.tool_calls, 1):
                        print(f"         å·¥å…·è°ƒç”¨ {j}: {tool_call.name}")
                        print(f"           å‚æ•°: {tool_call.args}")
                        
                        # æ‰§è¡Œå·¥å…·è°ƒç”¨
                        try:
                            # æ‰¾åˆ°å¯¹åº”çš„å·¥å…·
                            tool_to_call = None
                            for tool in tools_list:
                                if tool.name == tool_call.name:
                                    tool_to_call = tool
                                    break
                            
                            if tool_to_call:
                                tool_result = tool_to_call.invoke(tool_call.args)
                                print(f"           å·¥å…·æ‰§è¡Œç»“æœ: {tool_result}")
                                
                                gpt_test_results.append({
                                    "prompt": prompt,
                                    "gpt_response": response.content,
                                    "tool_called": tool_call.name,
                                    "tool_result": tool_result,
                                    "success": True
                                })
                            else:
                                print(f"           é”™è¯¯: æ‰¾ä¸åˆ°å·¥å…· {tool_call.name}")
                                gpt_test_results.append({
                                    "prompt": prompt,
                                    "success": False,
                                    "error": f"æ‰¾ä¸åˆ°å·¥å…· {tool_call.name}"
                                })
                        except Exception as e:
                            print(f"           å·¥å…·æ‰§è¡Œé”™è¯¯: {e}")
                            gpt_test_results.append({
                                "prompt": prompt,
                                "success": False,
                                "error": str(e)
                            })
                else:
                    print(f"       æœªæ£€æµ‹åˆ°å·¥å…·è°ƒç”¨ï¼Œç›´æ¥å“åº”")
                    gpt_test_results.append({
                        "prompt": prompt,
                        "gpt_response": response.content,
                        "success": True,
                        "direct_response": True
                    })
                    
            except Exception as e:
                print(f"       GPTè°ƒç”¨é”™è¯¯: {e}")
                gpt_test_results.append({
                    "prompt": prompt,
                    "success": False,
                    "error": str(e)
                })
        
        # 5. ç»“æœéªŒè¯
        print(f"\nâœ… 5. ç»“æœéªŒè¯:")
        
        def verify_gpt_tool_results(test_results):
            """éªŒè¯GPTå·¥å…·è°ƒç”¨ç»“æœ"""
            verification_stats = {
                "total_tests": len(test_results),
                "successful_tests": sum(1 for r in test_results if r.get("success", False)),
                "tool_calls_detected": sum(1 for r in test_results if "tool_called" in r),
                "direct_responses": sum(1 for r in test_results if r.get("direct_response", False)),
                "failed_tests": sum(1 for r in test_results if not r.get("success", False))
            }
            
            return verification_stats
        
        verification_stats = verify_gpt_tool_results(gpt_test_results)
        
        print(f"   GPTå·¥å…·è°ƒç”¨éªŒè¯ç»Ÿè®¡:")
        for stat_name, stat_value in verification_stats.items():
            print(f"     {stat_name}: {stat_value}")
        
        success_rate = verification_stats["successful_tests"] / verification_stats["total_tests"] * 100
        print(f"     æˆåŠŸç‡: {success_rate:.1f}%")
        
        # 6. è¯¦ç»†ç»“æœæ£€æŸ¥
        print(f"\nğŸ“Š 6. è¯¦ç»†ç»“æœæ£€æŸ¥:")
        
        print(f"   è¯¦ç»†æµ‹è¯•ç»“æœ:")
        for i, result in enumerate(gpt_test_results, 1):
            print(f"\n     æµ‹è¯• {i}: {result['prompt']}")
            if result.get("success", False):
                if "tool_called" in result:
                    print(f"       âœ… æˆåŠŸ - è°ƒç”¨å·¥å…·: {result['tool_called']}")
                    print(f"       å·¥å…·ç»“æœ: {result['tool_result']}")
                elif result.get("direct_response", False):
                    print(f"       âœ… æˆåŠŸ - ç›´æ¥å“åº”")
                    print(f"       å“åº”å†…å®¹: {result['gpt_response'][:50]}...")
            else:
                print(f"       âŒ å¤±è´¥ - {result.get('error', 'æœªçŸ¥é”™è¯¯')}")
        
        # 7. å·¥å…·æ‰§è¡Œè´¨é‡éªŒè¯
        print(f"\nğŸ¯ 7. å·¥å…·æ‰§è¡Œè´¨é‡éªŒè¯:")
        
        # æ£€æŸ¥å·¥å…·æ‰§è¡Œç»“æœçš„æ­£ç¡®æ€§
        quality_checks = []
        
        for result in gpt_test_results:
            if "tool_result" in result:
                tool_result = result["tool_result"]
                # æ£€æŸ¥ç»“æœæ˜¯å¦åŒ…å«é¢„æœŸçš„è®¡ç®—ç»“æœ
                if "+" in result["prompt"]:
                    expected = "42"  # 25 + 17 = 42
                    quality_checks.append({
                        "operation": "åŠ æ³•",
                        "expected": expected,
                        "actual": tool_result,
                        "correct": expected in tool_result
                    })
                elif "-" in result["prompt"]:
                    expected = "63"  # 100 - 37 = 63
                    quality_checks.append({
                        "operation": "å‡æ³•",
                        "expected": expected,
                        "actual": tool_result,
                        "correct": expected in tool_result
                    })
                elif "Ã—" in result["prompt"] or "*" in result["prompt"]:
                    expected = "72"  # 8 Ã— 9 = 72
                    quality_checks.append({
                        "operation": "ä¹˜æ³•",
                        "expected": expected,
                        "actual": tool_result,
                        "correct": expected in tool_result
                    })
                elif "Ã·" in result["prompt"] or "/" in result["prompt"]:
                    expected = "12"  # 144 Ã· 12 = 12
                    quality_checks.append({
                        "operation": "é™¤æ³•",
                        "expected": expected,
                        "actual": tool_result,
                        "correct": expected in tool_result
                    })
        
        print(f"   å·¥å…·æ‰§è¡Œè´¨é‡æ£€æŸ¥:")
        correct_calculations = 0
        for check in quality_checks:
            status = "âœ…" if check["correct"] else "âŒ"
            print(f"     {check['operation']}: {status} æœŸæœ›={check['expected']}, å®é™…={check['actual']}")
            if check["correct"]:
                correct_calculations += 1
        
        quality_rate = correct_calculations / len(quality_checks) * 100 if quality_checks else 0
        print(f"\n   è®¡ç®—æ­£ç¡®ç‡: {quality_rate:.1f}%")
        
        # ä¿å­˜ç»“æœä¾›åç»­ä½¿ç”¨
        globals().update({
            'gpt_test_results': gpt_test_results,
            'llm_with_tools': llm_with_tools,
            'quality_checks': quality_checks
        })
        
        # éªŒè¯ç‚¹ï¼šå·¥å…·èƒ½æ­£ç¡®æ‰§è¡Œå¹¶è¿”å›GPTç»“æœ
        assert verification_stats["total_tests"] == 4, "åº”è¯¥æœ‰4ä¸ªæµ‹è¯•æ¡ˆä¾‹"
        assert verification_stats["successful_tests"] >= 2, "è‡³å°‘åº”è¯¥æœ‰2ä¸ªæµ‹è¯•æˆåŠŸ"
        assert success_rate >= 50, "æˆåŠŸç‡åº”è¯¥è‡³å°‘50%"
        assert len(quality_checks) > 0, "åº”è¯¥æœ‰è´¨é‡æ£€æŸ¥ç»“æœ"
        
        print(f"\nâœ… éªŒè¯é€šè¿‡ï¼šå·¥å…·èƒ½æ­£ç¡®æ‰§è¡Œå¹¶è¿”å›GPTç»“æœ")
        print(f"\nğŸ¯ GPTå·¥å…·è°ƒç”¨æ€»ç»“:")
        print(f"   âœ“ å·¥å…·æˆåŠŸç»‘å®šåˆ°GPT LLM")
        print(f"   âœ“ GPTèƒ½å¤Ÿè¯†åˆ«å¹¶è°ƒç”¨è®¡ç®—å·¥å…·")
        print(f"   âœ“ å·¥å…·æ‰§è¡Œç»“æœæ­£ç¡®è¿”å›")
        print(f"   âœ“ è®¡ç®—ç»“æœéªŒè¯æ­£ç¡®")
        print(f"   âœ“ å®Œæ•´çš„å·¥å…·è°ƒç”¨æµç¨‹éªŒè¯")
        
    except Exception as e:
        print(f"\nâŒ éªŒè¯å·¥å…·èƒ½æ­£ç¡®æ‰§è¡Œå¹¶è¿”å›GPTç»“æœå¤±è´¥: {e}")
```
