### batch

```python
from datetime import datetime

import dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

# 1.ç¼–æ’prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "ä½ æ˜¯OpenAIå¼€å‘çš„èŠå¤©æœºå™¨äººï¼Œè¯·å›ç­”ç”¨æˆ·çš„é—®é¢˜ï¼Œç°åœ¨çš„æ—¶é—´æ˜¯{now}"),
    ("human", "{query}"),
]).partial(now=datetime.now())

# 2.åˆ›å»ºå¤§è¯­è¨€æ¨¡å‹
llm = ChatOpenAI(
        model="gpt-4o",  # ä½¿ç”¨ GPT-4o æ¨¡å‹è¿›è¡Œæµ‹è¯•
        temperature=0.1,
        max_tokens=50
    )

ai_messages = llm.batch([
    prompt.invoke({"query": "ä½ å¥½ï¼Œä½ æ˜¯?"}),
    prompt.invoke({"query": "è¯·è®²ä¸€ä¸ªå…³äºç¨‹åºå‘˜çš„å†·ç¬‘è¯"}),
])

for ai_message in ai_messages:
    print(ai_message.content)
    print("====================")
```

```python
from datetime import datetime

import dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

# 1.ç¼–æ’prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "ä½ æ˜¯OpenAIå¼€å‘çš„èŠå¤©æœºå™¨äººï¼Œè¯·å›ç­”ç”¨æˆ·çš„é—®é¢˜ï¼Œç°åœ¨çš„æ—¶é—´æ˜¯{now}"),
    ("human", "{query}"),
]).partial(now=datetime.now())

# 2.åˆ›å»ºå¤§è¯­è¨€æ¨¡å‹
llm = ChatOpenAI(
        model="gpt-4o",  # ä½¿ç”¨ GPT-4o æ¨¡å‹è¿›è¡Œæµ‹è¯•
        temperature=0.1,
        max_tokens=200
    )

response = llm.stream(prompt.invoke({"query": "ä½ èƒ½ç®€å•ä»‹ç»ä¸‹LLMå’ŒLLMOpså—?"}))

for chunk in response:
    print(chunk.content, flush=True, end="")
```
