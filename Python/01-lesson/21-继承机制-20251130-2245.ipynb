````python
print("ğŸ§¬ ç»§æ‰¿æœºåˆ¶åŸºç¡€:")
print("=" * 50)

# 1. å•ç»§æ‰¿åŸºç¡€
print(f"ğŸ“ 1. å•ç»§æ‰¿åŸºç¡€:")

class Animal:
    """åŠ¨ç‰©åŸºç±»"""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"   åˆ›å»ºåŠ¨ç‰©: {name}, {age}å²")
    
    def eat(self):
        """åƒä¸œè¥¿"""
        return f"{self.name} æ­£åœ¨åƒä¸œè¥¿"
    
    def sleep(self):
        """ç¡è§‰"""
        return f"{self.name} æ­£åœ¨ç¡è§‰"
    
    def make_sound(self):
        """å‘å‡ºå£°éŸ³"""
        return f"{self.name} å‘å‡ºå£°éŸ³"
    
    def get_info(self):
        """è·å–ä¿¡æ¯"""
        return f"åŠ¨ç‰©: {self.name}, {self.age}å²"

class Dog(Animal):
    """ç‹—ç±»ï¼šç»§æ‰¿è‡ªåŠ¨ç‰©ç±»"""
    
    def __init__(self, name, age, breed):
        # è°ƒç”¨çˆ¶ç±»æ„é€ å‡½æ•°
        super().__init__(name, age)
        self.breed = breed
        print(f"   åˆ›å»ºç‹—: {name}, å“ç§: {breed}")
    
    def make_sound(self):
        """é‡å†™çˆ¶ç±»æ–¹æ³•"""
        return f"{self.name} æ±ªæ±ªå«"
    
    def wag_tail(self):
        """å­ç±»ç‰¹æœ‰æ–¹æ³•"""
        return f"{self.name} æ‘‡å°¾å·´"
    
    def fetch(self):
        """å­ç±»ç‰¹æœ‰æ–¹æ³•"""
        return f"{self.name} å»æ¡çƒ"
    
    def get_info(self):
        """é‡å†™çˆ¶ç±»æ–¹æ³•å¹¶è°ƒç”¨çˆ¶ç±»æ–¹æ³•"""
        base_info = super().get_info()
        return f"{base_info}, å“ç§: {self.breed}"

class Cat(Animal):
    """çŒ«ç±»ï¼šç»§æ‰¿è‡ªåŠ¨ç‰©ç±»"""
    
    def __init__(self, name, age, color):
        super().__init__(name, age)
        self.color = color
        print(f"   åˆ›å»ºçŒ«: {name}, é¢œè‰²: {color}")
    
    def make_sound(self):
        """é‡å†™çˆ¶ç±»æ–¹æ³•"""
        return f"{self.name} å–µå–µå«"
    
    def climb(self):
        """å­ç±»ç‰¹æœ‰æ–¹æ³•"""
        return f"{self.name} çˆ¬æ ‘"
    
    def get_info(self):
        """é‡å†™çˆ¶ç±»æ–¹æ³•"""
        return f"çŒ«: {self.name}, {self.age}å², é¢œè‰²: {self.color}"

# æµ‹è¯•å•ç»§æ‰¿
print(f"   åˆ›å»ºåŠ¨ç‰©å¯¹è±¡:")
animal = Animal("é€šç”¨åŠ¨ç‰©", 5)
print(f"     {animal.get_info()}")
print(f"     {animal.eat()}")
print(f"     {animal.make_sound()}")

print(f"\n   åˆ›å»ºç‹—å¯¹è±¡:")
dog = Dog("æ—ºè´¢", 3, "é‡‘æ¯›")
print(f"     {dog.get_info()}")
print(f"     {dog.eat()}")        # ç»§æ‰¿çš„æ–¹æ³•
print(f"     {dog.make_sound()}") # é‡å†™çš„æ–¹æ³•
print(f"     {dog.wag_tail()}")   # å­ç±»ç‰¹æœ‰æ–¹æ³•

print(f"\n   åˆ›å»ºçŒ«å¯¹è±¡:")
cat = Cat("å’ªå’ª", 2, "æ©˜è‰²")
print(f"     {cat.get_info()}")
print(f"     {cat.sleep()}")      # ç»§æ‰¿çš„æ–¹æ³•
print(f"     {cat.make_sound()}") # é‡å†™çš„æ–¹æ³•
print(f"     {cat.climb()}")      # å­ç±»ç‰¹æœ‰æ–¹æ³•

# 2. æ–¹æ³•é‡å†™è¯¦è§£
print(f"\nğŸ“ 2. æ–¹æ³•é‡å†™è¯¦è§£:")

class Vehicle:
    """äº¤é€šå·¥å…·åŸºç±»"""
    
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.speed = 0
    
    def start(self):
        """å¯åŠ¨"""
        return f"{self.brand} {self.model} å¯åŠ¨äº†"
    
    def stop(self):
        """åœæ­¢"""
        self.speed = 0
        return f"{self.brand} {self.model} åœæ­¢äº†"
    
    def accelerate(self, amount):
        """åŠ é€Ÿ"""
        self.speed += amount
        return f"{self.brand} {self.model} åŠ é€Ÿåˆ° {self.speed} km/h"
    
    def brake(self, amount):
        """åˆ¹è½¦"""
        self.speed = max(0, self.speed - amount)
        return f"{self.brand} {self.model} å‡é€Ÿåˆ° {self.speed} km/h"

class Car(Vehicle):
    """æ±½è½¦ç±»"""
    
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)
        self.doors = doors
        self.fuel_type = "æ±½æ²¹"
    
    def start(self):
        """é‡å†™å¯åŠ¨æ–¹æ³•"""
        base_start = super().start()
        return f"{base_start}ï¼Œå¼•æ“è½°é¸£å£°"
    
    def open_trunk(self):
        """æ‰“å¼€åå¤‡ç®±"""
        return f"{self.brand} {self.model} åå¤‡ç®±å·²æ‰“å¼€"

class ElectricCar(Car):
    """ç”µåŠ¨æ±½è½¦ç±»"""
    
    def __init__(self, brand, model, doors, battery_capacity):
        super().__init__(brand, model, doors)
        self.battery_capacity = battery_capacity
        self.fuel_type = "ç”µåŠ›"
        self.battery_level = 100
    
    def start(self):
        """é‡å†™å¯åŠ¨æ–¹æ³•"""
        return f"{self.brand} {self.model} é™éŸ³å¯åŠ¨ï¼Œç”µæ± ç”µé‡: {self.battery_level}%"
    
    def charge(self, amount):
        """å……ç”µ"""
        self.battery_level = min(100, self.battery_level + amount)
        return f"å……ç”µä¸­... å½“å‰ç”µé‡: {self.battery_level}%"
    
    def get_range(self):
        """è·å–ç»­èˆªé‡Œç¨‹"""
        range_km = self.battery_level * self.battery_capacity / 100
        return f"ç»­èˆªé‡Œç¨‹: {range_km:.0f} km"

# æµ‹è¯•æ–¹æ³•é‡å†™
print(f"   æ–¹æ³•é‡å†™æµ‹è¯•:")
car = Car("ä¸°ç”°", "å‡¯ç¾ç‘", 4)
print(f"     {car.start()}")
print(f"     {car.accelerate(60)}")
print(f"     {car.open_trunk()}")

tesla = ElectricCar("ç‰¹æ–¯æ‹‰", "Model 3", 4, 400)
print(f"\n     {tesla.start()}")
print(f"     {tesla.accelerate(80)}")
print(f"     {tesla.get_range()}")
print(f"     {tesla.charge(20)}")
print(f"     {tesla.get_range()}")

# 3. super()å‡½æ•°è¯¦è§£
print(f"\nğŸ“ 3. super()å‡½æ•°è¯¦è§£:")

class Person:
    """äººç±»åŸºç±»"""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"   Person.__init__: {name}, {age}")
    
    def introduce(self):
        return f"æˆ‘å«{self.name}ï¼Œä»Šå¹´{self.age}å²"
    
    def work(self):
        return f"{self.name} æ­£åœ¨å·¥ä½œ"

class Employee(Person):
    """å‘˜å·¥ç±»"""
    
    def __init__(self, name, age, employee_id, salary):
        # æ–¹å¼1ï¼šç›´æ¥è°ƒç”¨çˆ¶ç±»æ„é€ å‡½æ•°
        Person.__init__(self, name, age)
        self.employee_id = employee_id
        self.salary = salary
        print(f"   Employee.__init__: {employee_id}, {salary}")
    
    def introduce(self):
        # æ–¹å¼2ï¼šä½¿ç”¨super()è°ƒç”¨çˆ¶ç±»æ–¹æ³•
        base_intro = super().introduce()
        return f"{base_intro}ï¼Œå‘˜å·¥ç¼–å·: {self.employee_id}"
    
    def work(self):
        base_work = super().work()
        return f"{base_work}ï¼Œæœˆè–ª: {self.salary}"

class Manager(Employee):
    """ç»ç†ç±»"""
    
    def __init__(self, name, age, employee_id, salary, department):
        # æ–¹å¼3ï¼šä½¿ç”¨super()è°ƒç”¨çˆ¶ç±»æ„é€ å‡½æ•°
        super().__init__(name, age, employee_id, salary)
        self.department = department
        self.team = []
        print(f"   Manager.__init__: {department}")
    
    def introduce(self):
        base_intro = super().introduce()
        return f"{base_intro}ï¼Œéƒ¨é—¨: {self.department}ç»ç†"
    
    def add_team_member(self, employee):
        """æ·»åŠ å›¢é˜Ÿæˆå‘˜"""
        self.team.append(employee)
        print(f"   ç»ç†{self.name}æ·»åŠ äº†å›¢é˜Ÿæˆå‘˜: {employee.name}")
    
    def manage_team(self):
        """ç®¡ç†å›¢é˜Ÿ"""
        team_names = [emp.name for emp in self.team]
        return f"ç»ç†{self.name}ç®¡ç†å›¢é˜Ÿ: {team_names}"

# æµ‹è¯•super()å‡½æ•°
print(f"   super()å‡½æ•°æµ‹è¯•:")
employee = Employee("å¼ ä¸‰", 28, "E001", 8000)
print(f"     {employee.introduce()}")
print(f"     {employee.work()}")

manager = Manager("æå››", 35, "M001", 15000, "æŠ€æœ¯éƒ¨")
print(f"\n     {manager.introduce()}")
print(f"     {manager.work()}")

# æ·»åŠ å›¢é˜Ÿæˆå‘˜
manager.add_team_member(employee)
employee2 = Employee("ç‹äº”", 25, "E002", 7000)
manager.add_team_member(employee2)
print(f"     {manager.manage_team()}")

# 4. isinstance()å’Œissubclass()
print(f"\nğŸ“ 4. isinstance()å’Œissubclass():")

# ç±»å‹æ£€æŸ¥
print(f"   isinstance()æ£€æŸ¥:")
print(f"     dogæ˜¯Dogçš„å®ä¾‹: {isinstance(dog, Dog)}")
print(f"     dogæ˜¯Animalçš„å®ä¾‹: {isinstance(dog, Animal)}")
print(f"     dogæ˜¯Catçš„å®ä¾‹: {isinstance(dog, Cat)}")
print(f"     catæ˜¯Animalçš„å®ä¾‹: {isinstance(cat, Animal)}")
print(f"     teslaæ˜¯ElectricCarçš„å®ä¾‹: {isinstance(tesla, ElectricCar)}")
print(f"     teslaæ˜¯Carçš„å®ä¾‹: {isinstance(tesla, Car)}")
print(f"     teslaæ˜¯Vehicleçš„å®ä¾‹: {isinstance(tesla, Vehicle)}")

print(f"\n   issubclass()æ£€æŸ¥:")
print(f"     Dogæ˜¯Animalçš„å­ç±»: {issubclass(Dog, Animal)}")
print(f"     Catæ˜¯Animalçš„å­ç±»: {issubclass(Cat, Animal)}")
print(f"     Animalæ˜¯Dogçš„å­ç±»: {issubclass(Animal, Dog)}")
print(f"     ElectricCaræ˜¯Carçš„å­ç±»: {issubclass(ElectricCar, Car)}")
print(f"     ElectricCaræ˜¯Vehicleçš„å­ç±»: {issubclass(ElectricCar, Vehicle)}")

# 5. ç»§æ‰¿å±‚æ¬¡è®¾è®¡åŸåˆ™
print(f"\nğŸ“ 5. ç»§æ‰¿å±‚æ¬¡è®¾è®¡åŸåˆ™:")

print(f"   ç»§æ‰¿è®¾è®¡åŸåˆ™:")
print(f"     - IS-Aå…³ç³»ï¼šå­ç±»åº”è¯¥æ˜¯çˆ¶ç±»çš„ä¸€ç§ç‰¹æ®Šç±»å‹")
print(f"     - é‡Œæ°æ›¿æ¢åŸåˆ™ï¼šå­ç±»åº”è¯¥èƒ½å¤Ÿæ›¿æ¢çˆ¶ç±»")
print(f"     - å•ä¸€èŒè´£ï¼šæ¯ä¸ªç±»åº”è¯¥æœ‰æ˜ç¡®çš„èŒè´£")
print(f"     - é¿å…è¿‡æ·±çš„ç»§æ‰¿å±‚æ¬¡")
print(f"     - ä¼˜å…ˆä½¿ç”¨ç»„åˆè€Œä¸æ˜¯ç»§æ‰¿")

print(f"\n   è®¾è®¡æ¨¡å¼å»ºè®®:")
print(f"     - æ¨¡æ¿æ–¹æ³•æ¨¡å¼ï¼šåœ¨åŸºç±»ä¸­å®šä¹‰ç®—æ³•æ¡†æ¶")
print(f"     - ç­–ç•¥æ¨¡å¼ï¼šé€šè¿‡ç»§æ‰¿å®ç°ä¸åŒçš„ç­–ç•¥")
print(f"     - å·¥å‚æ–¹æ³•æ¨¡å¼ï¼šé€šè¿‡å­ç±»åˆ›å»ºå…·ä½“äº§å“")

print(f"\nâœ… ç»§æ‰¿æœºåˆ¶åŸºç¡€å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡ç»§æ‰¿çš„åŸºæœ¬è¯­æ³•å’Œç”¨æ³•")
print(f"   âœ“ ç†è§£æ–¹æ³•é‡å†™å’Œsuper()å‡½æ•°çš„ä½¿ç”¨")
print(f"   âœ“ èƒ½è®¾è®¡åˆç†çš„ç±»å±‚æ¬¡ç»“æ„")
````

````python
print("ğŸ—ï¸ å®é™…åº”ç”¨æ¡ˆä¾‹:")
print("=" * 50)

# æ¡ˆä¾‹1: å›¾å½¢ç»˜åˆ¶ç³»ç»Ÿ
print(f"ğŸ“ 1. å›¾å½¢ç»˜åˆ¶ç³»ç»Ÿ:")

class Shape:
    """å›¾å½¢åŸºç±»"""
    
    def __init__(self, x, y, color="black"):
        self.x = x
        self.y = y
        self.color = color
    
    def area(self):
        """è®¡ç®—é¢ç§¯"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°areaæ–¹æ³•")
    
    def perimeter(self):
        """è®¡ç®—å‘¨é•¿"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°perimeteræ–¹æ³•")
    
    def draw(self):
        """ç»˜åˆ¶å›¾å½¢"""
        return f"åœ¨({self.x}, {self.y})ä½ç½®ç»˜åˆ¶{self.color}è‰²çš„{self.__class__.__name__}"
    
    def move(self, dx, dy):
        """ç§»åŠ¨å›¾å½¢"""
        self.x += dx
        self.y += dy
        return f"å›¾å½¢ç§»åŠ¨åˆ°({self.x}, {self.y})"

class Circle(Shape):
    """åœ†å½¢ç±»"""
    
    def __init__(self, x, y, radius, color="black"):
        super().__init__(x, y, color)
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
    def perimeter(self):
        return 2 * 3.14159 * self.radius
    
    def draw(self):
        base_draw = super().draw()
        return f"{base_draw}ï¼ŒåŠå¾„: {self.radius}"

class Rectangle(Shape):
    """çŸ©å½¢ç±»"""
    
    def __init__(self, x, y, width, height, color="black"):
        super().__init__(x, y, color)
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def draw(self):
        base_draw = super().draw()
        return f"{base_draw}ï¼Œå®½: {self.width}, é«˜: {self.height}"

class Triangle(Shape):
    """ä¸‰è§’å½¢ç±»"""
    
    def __init__(self, x, y, side1, side2, side3, color="black"):
        super().__init__(x, y, color)
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
    
    def area(self):
        """ä½¿ç”¨æµ·ä¼¦å…¬å¼è®¡ç®—é¢ç§¯"""
        s = self.perimeter() / 2
        return (s * (s - self.side1) * (s - self.side2) * (s - self.side3)) ** 0.5
    
    def perimeter(self):
        return self.side1 + self.side2 + self.side3
    
    def draw(self):
        base_draw = super().draw()
        return f"{base_draw}ï¼Œè¾¹é•¿: {self.side1}, {self.side2}, {self.side3}"

class Square(Rectangle):
    """æ­£æ–¹å½¢ç±»ï¼šç»§æ‰¿è‡ªçŸ©å½¢"""
    
    def __init__(self, x, y, side, color="black"):
        super().__init__(x, y, side, side, color)
    
    def draw(self):
        base_draw = super().draw()
        return f"{base_draw}ï¼ˆæ­£æ–¹å½¢ï¼‰"

# åˆ›å»ºå›¾å½¢å¯¹è±¡
shapes = [
    Circle(10, 10, 5, "red"),
    Rectangle(20, 20, 8, 6, "blue"),
    Triangle(30, 30, 3, 4, 5, "green"),
    Square(40, 40, 7, "yellow")
]

print(f"   å›¾å½¢ç»˜åˆ¶ç³»ç»Ÿæµ‹è¯•:")
for shape in shapes:
    print(f"\n     {shape.draw()}")
    print(f"       é¢ç§¯: {shape.area():.2f}")
    print(f"       å‘¨é•¿: {shape.perimeter():.2f}")
    shape.move(5, 5)
    print(f"       ç§»åŠ¨å: {shape.draw()}")

# æ¡ˆä¾‹2: æ–‡ä»¶å¤„ç†ç³»ç»Ÿ
print(f"\nğŸ“ 2. æ–‡ä»¶å¤„ç†ç³»ç»Ÿ:")

class FileProcessor:
    """æ–‡ä»¶å¤„ç†å™¨åŸºç±»"""
    
    def __init__(self, filename):
        self.filename = filename
        self.content = ""
        self.is_loaded = False
    
    def load(self):
        """åŠ è½½æ–‡ä»¶"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°loadæ–¹æ³•")
    
    def save(self):
        """ä¿å­˜æ–‡ä»¶"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°saveæ–¹æ³•")
    
    def process(self):
        """å¤„ç†æ–‡ä»¶"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°processæ–¹æ³•")
    
    def get_info(self):
        return f"æ–‡ä»¶: {self.filename}, å·²åŠ è½½: {self.is_loaded}"

class TextFileProcessor(FileProcessor):
    """æ–‡æœ¬æ–‡ä»¶å¤„ç†å™¨"""
    
    def load(self):
        """åŠ è½½æ–‡æœ¬æ–‡ä»¶"""
        # æ¨¡æ‹ŸåŠ è½½è¿‡ç¨‹
        self.content = "è¿™æ˜¯ä¸€ä¸ªæ–‡æœ¬æ–‡ä»¶çš„å†…å®¹ã€‚\nåŒ…å«å¤šè¡Œæ–‡æœ¬ã€‚"
        self.is_loaded = True
        print(f"   æ–‡æœ¬æ–‡ä»¶ {self.filename} åŠ è½½æˆåŠŸ")
    
    def save(self):
        """ä¿å­˜æ–‡æœ¬æ–‡ä»¶"""
        if self.is_loaded:
            print(f"   æ–‡æœ¬æ–‡ä»¶ {self.filename} ä¿å­˜æˆåŠŸ")
            return True
        return False
    
    def process(self):
        """å¤„ç†æ–‡æœ¬æ–‡ä»¶"""
        if not self.is_loaded:
            self.load()
        
        # ç»Ÿè®¡ä¿¡æ¯
        lines = len(self.content.split('\n'))
        words = len(self.content.split())
        chars = len(self.content)
        
        return {
            "lines": lines,
            "words": words,
            "characters": chars
        }

class JSONFileProcessor(FileProcessor):
    """JSONæ–‡ä»¶å¤„ç†å™¨"""
    
    def load(self):
        """åŠ è½½JSONæ–‡ä»¶"""
        # æ¨¡æ‹ŸåŠ è½½JSONæ•°æ®
        self.content = '{"name": "å¼ ä¸‰", "age": 25, "city": "åŒ—äº¬"}'
        self.is_loaded = True
        print(f"   JSONæ–‡ä»¶ {self.filename} åŠ è½½æˆåŠŸ")
    
    def save(self):
        """ä¿å­˜JSONæ–‡ä»¶"""
        if self.is_loaded:
            print(f"   JSONæ–‡ä»¶ {self.filename} ä¿å­˜æˆåŠŸ")
            return True
        return False
    
    def process(self):
        """å¤„ç†JSONæ–‡ä»¶"""
        if not self.is_loaded:
            self.load()
        
        # è§£æJSONï¼ˆç®€åŒ–ç‰ˆï¼‰
        try:
            import json
            data = json.loads(self.content)
            return {
                "keys": list(data.keys()),
                "values": list(data.values()),
                "type": "JSON"
            }
        except:
            return {"error": "JSONè§£æå¤±è´¥"}

class CSVFileProcessor(FileProcessor):
    """CSVæ–‡ä»¶å¤„ç†å™¨"""
    
    def load(self):
        """åŠ è½½CSVæ–‡ä»¶"""
        # æ¨¡æ‹ŸåŠ è½½CSVæ•°æ®
        self.content = "å§“å,å¹´é¾„,åŸå¸‚\nå¼ ä¸‰,25,åŒ—äº¬\næå››,30,ä¸Šæµ·"
        self.is_loaded = True
        print(f"   CSVæ–‡ä»¶ {self.filename} åŠ è½½æˆåŠŸ")
    
    def save(self):
        """ä¿å­˜CSVæ–‡ä»¶"""
        if self.is_loaded:
            print(f"   CSVæ–‡ä»¶ {self.filename} ä¿å­˜æˆåŠŸ")
            return True
        return False
    
    def process(self):
        """å¤„ç†CSVæ–‡ä»¶"""
        if not self.is_loaded:
            self.load()
        
        # è§£æCSVæ•°æ®
        lines = self.content.split('\n')
        headers = lines[0].split(',')
        data = []
        
        for line in lines[1:]:
            if line.strip():
                values = line.split(',')
                row = dict(zip(headers, values))
                data.append(row)
        
        return {
            "headers": headers,
            "rows": len(data),
            "data": data,
            "type": "CSV"
        }

# æµ‹è¯•æ–‡ä»¶å¤„ç†ç³»ç»Ÿ
processors = [
    TextFileProcessor("document.txt"),
    JSONFileProcessor("data.json"),
    CSVFileProcessor("users.csv")
]

print(f"\n   æ–‡ä»¶å¤„ç†ç³»ç»Ÿæµ‹è¯•:")
for processor in processors:
    print(f"\n     {processor.get_info()}")
    result = processor.process()
    print(f"       å¤„ç†ç»“æœ: {result}")
    processor.save()

# æ¡ˆä¾‹3: æ”¯ä»˜ç³»ç»Ÿ
print(f"\nğŸ“ 3. æ”¯ä»˜ç³»ç»Ÿ:")

class PaymentMethod:
    """æ”¯ä»˜æ–¹å¼åŸºç±»"""
    
    def __init__(self, account_id, balance=0):
        self.account_id = account_id
        self.balance = balance
        self.transaction_history = []
    
    def pay(self, amount):
        """æ”¯ä»˜"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°payæ–¹æ³•")
    
    def refund(self, amount):
        """é€€æ¬¾"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°refundæ–¹æ³•")
    
    def get_balance(self):
        return self.balance
    
    def add_transaction(self, transaction_type, amount):
        self.transaction_history.append({
            "type": transaction_type,
            "amount": amount,
            "balance_after": self.balance
        })

class CreditCard(PaymentMethod):
    """ä¿¡ç”¨å¡æ”¯ä»˜"""
    
    def __init__(self, card_number, cvv, expiry_date, credit_limit):
        super().__init__(card_number)
        self.cvv = cvv
        self.expiry_date = expiry_date
        self.credit_limit = credit_limit
        self.used_credit = 0
    
    def pay(self, amount):
        """ä¿¡ç”¨å¡æ”¯ä»˜"""
        if self.used_credit + amount <= self.credit_limit:
            self.used_credit += amount
            self.add_transaction("æ”¯ä»˜", amount)
            print(f"   ä¿¡ç”¨å¡æ”¯ä»˜æˆåŠŸ: {amount}")
            return True
        else:
            print(f"   ä¿¡ç”¨é¢åº¦ä¸è¶³")
            return False
    
    def refund(self, amount):
        """ä¿¡ç”¨å¡é€€æ¬¾"""
        if self.used_credit >= amount:
            self.used_credit -= amount
            self.add_transaction("é€€æ¬¾", -amount)
            print(f"   ä¿¡ç”¨å¡é€€æ¬¾æˆåŠŸ: {amount}")
            return True
        return False
    
    def get_available_credit(self):
        return self.credit_limit - self.used_credit

class DebitCard(PaymentMethod):
    """å€Ÿè®°å¡æ”¯ä»˜"""
    
    def __init__(self, card_number, cvv, expiry_date, account_balance):
        super().__init__(card_number, account_balance)
        self.cvv = cvv
        self.expiry_date = expiry_date
    
    def pay(self, amount):
        """å€Ÿè®°å¡æ”¯ä»˜"""
        if self.balance >= amount:
            self.balance -= amount
            self.add_transaction("æ”¯ä»˜", amount)
            print(f"   å€Ÿè®°å¡æ”¯ä»˜æˆåŠŸ: {amount}")
            return True
        else:
            print(f"   ä½™é¢ä¸è¶³")
            return False
    
    def refund(self, amount):
        """å€Ÿè®°å¡é€€æ¬¾"""
        self.balance += amount
        self.add_transaction("é€€æ¬¾", -amount)
        print(f"   å€Ÿè®°å¡é€€æ¬¾æˆåŠŸ: {amount}")
        return True

class DigitalWallet(PaymentMethod):
    """æ•°å­—é’±åŒ…æ”¯ä»˜"""
    
    def __init__(self, wallet_id, phone_number, balance=0):
        super().__init__(wallet_id, balance)
        self.phone_number = phone_number
        self.daily_limit = 5000
        self.daily_spent = 0
    
    def pay(self, amount):
        """æ•°å­—é’±åŒ…æ”¯ä»˜"""
        if self.balance >= amount and self.daily_spent + amount <= self.daily_limit:
            self.balance -= amount
            self.daily_spent += amount
            self.add_transaction("æ”¯ä»˜", amount)
            print(f"   æ•°å­—é’±åŒ…æ”¯ä»˜æˆåŠŸ: {amount}")
            return True
        elif self.daily_spent + amount > self.daily_limit:
            print(f"   è¶…å‡ºæ¯æ—¥é™é¢ {self.daily_limit}")
            return False
        else:
            print(f"   ä½™é¢ä¸è¶³")
            return False
    
    def refund(self, amount):
        """æ•°å­—é’±åŒ…é€€æ¬¾"""
        self.balance += amount
        self.daily_spent = max(0, self.daily_spent - amount)
        self.add_transaction("é€€æ¬¾", -amount)
        print(f"   æ•°å­—é’±åŒ…é€€æ¬¾æˆåŠŸ: {amount}")
        return True
    
    def reset_daily_limit(self):
        """é‡ç½®æ¯æ—¥é™é¢"""
        self.daily_spent = 0
        print(f"   æ¯æ—¥é™é¢å·²é‡ç½®")

# æµ‹è¯•æ”¯ä»˜ç³»ç»Ÿ
print(f"\n   æ”¯ä»˜ç³»ç»Ÿæµ‹è¯•:")

credit_card = CreditCard("1234567812345678", "123", "2025-12", 10000)
debit_card = DebitCard("8765432187654321", "456", "2025-12", 5000)
wallet = DigitalWallet("WALLET001", "13800138000", 2000)

payments = [
    ("ä¿¡ç”¨å¡", credit_card, 3000),
    ("å€Ÿè®°å¡", debit_card, 1000),
    ("æ•°å­—é’±åŒ…", wallet, 800)
]

for name, payment_method, amount in payments:
    print(f"\n     {name}æ”¯ä»˜æµ‹è¯•:")
    payment_method.pay(amount)
    print(f"       å½“å‰çŠ¶æ€: {payment_method.get_balance()}")
    payment_method.refund(200)
    print(f"       é€€æ¬¾å: {payment_method.get_balance()}")

print(f"\nâœ… å®é™…åº”ç”¨æ¡ˆä¾‹å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡ç»§æ‰¿åœ¨å®é™…ç³»ç»Ÿä¸­çš„åº”ç”¨")
print(f"   âœ“ ç†è§£ç»§æ‰¿ä¸ç»„åˆçš„é€‰æ‹©åŸåˆ™")
print(f"   âœ“ èƒ½åº”ç”¨ç»§æ‰¿è§£å†³å¤æ‚ä¸šåŠ¡é—®é¢˜")
````
