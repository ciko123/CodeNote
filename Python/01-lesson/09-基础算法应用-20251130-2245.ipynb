### 1.9 åŸºç¡€ç®—æ³•åº”ç”¨ [â­â­è¿›é˜¶]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šåŸºç¡€ç®—æ³•åŒ…æ‹¬æ’åºã€æœç´¢ã€é€’å½’ç­‰å¸¸è§é—®é¢˜è§£å†³æ–¹æ³•ï¼Œæ˜¯ç¼–ç¨‹æ€ç»´çš„é‡è¦ä½“ç°ã€‚

**å­¦ä¹ è¦æ±‚**ï¼š
- æŒæ¡åŸºç¡€æ’åºç®—æ³•å®ç°
- ç†è§£æœç´¢ç®—æ³•çš„åŸç†
- èƒ½è¿›è¡Œé€’å½’é—®é¢˜çš„è§£å†³

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°åŸºç¡€æ’åºç®—æ³•
- è¿›è¡Œæœç´¢ç®—æ³•åº”ç”¨
- åº”ç”¨é€’å½’è§£å†³å®é™…é—®é¢˜
- éªŒè¯ç‚¹ï¼šèƒ½æ­£ç¡®å®ç°åŸºç¡€ç®—æ³•

````python
print("ğŸ”¢ åŸºç¡€æ’åºç®—æ³•:")
print("=" * 50)

# å†’æ³¡æ’åº
print(f"ğŸ“ 1. å†’æ³¡æ’åº:")
def bubble_sort(arr):
    """å†’æ³¡æ’åºç®—æ³•"""
    n = len(arr)
    arr = arr.copy()  # é¿å…ä¿®æ”¹åŸæ•°ç»„
    
    for i in range(n):
        # æ¯è½®å°†æœ€å¤§çš„å…ƒç´ å†’æ³¡åˆ°æœ€å
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
    return arr

# æµ‹è¯•å†’æ³¡æ’åº
test_data = [64, 34, 25, 12, 22, 11, 90]
print(f"   åŸå§‹æ•°æ®: {test_data}")
sorted_data = bubble_sort(test_data)
print(f"   å†’æ³¡æ’åºç»“æœ: {sorted_data}")

# é€‰æ‹©æ’åº
print(f"\nğŸ“ 2. é€‰æ‹©æ’åº:")
def selection_sort(arr):
    """é€‰æ‹©æ’åºç®—æ³•"""
    n = len(arr)
    arr = arr.copy()
    
    for i in range(n):
        # æ‰¾åˆ°æœ€å°å…ƒç´ çš„ç´¢å¼•
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        # å°†æœ€å°å…ƒç´ äº¤æ¢åˆ°å½“å‰ä½ç½®
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
    return arr

# æµ‹è¯•é€‰æ‹©æ’åº
sorted_data2 = selection_sort(test_data)
print(f"   é€‰æ‹©æ’åºç»“æœ: {sorted_data2}")

# æ’å…¥æ’åº
print(f"\nğŸ“ 3. æ’å…¥æ’åº:")
def insertion_sort(arr):
    """æ’å…¥æ’åºç®—æ³•"""
    arr = arr.copy()
    
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        # å°†å¤§äºkeyçš„å…ƒç´ å‘åç§»åŠ¨
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        
        arr[j + 1] = key
    
    return arr

# æµ‹è¯•æ’å…¥æ’åº
sorted_data3 = insertion_sort(test_data)
print(f"   æ’å…¥æ’åºç»“æœ: {sorted_data3}")

# å¿«é€Ÿæ’åº
print(f"\nğŸ“ 4. å¿«é€Ÿæ’åº:")
def quick_sort(arr):
    """å¿«é€Ÿæ’åºç®—æ³•"""
    if len(arr) <= 1:
        return arr
    
    # é€‰æ‹©åŸºå‡†å…ƒç´ 
    pivot = arr[len(arr) // 2]
    
    # åˆ†å‰²æ•°ç»„
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    # é€’å½’æ’åº
    return quick_sort(left) + middle + quick_sort(right)

# æµ‹è¯•å¿«é€Ÿæ’åº
sorted_data4 = quick_sort(test_data)
print(f"   å¿«é€Ÿæ’åºç»“æœ: {sorted_data4}")

# å½’å¹¶æ’åº
print(f"\nğŸ“ 5. å½’å¹¶æ’åº:")
def merge_sort(arr):
    """å½’å¹¶æ’åºç®—æ³•"""
    if len(arr) <= 1:
        return arr
    
    # åˆ†å‰²æ•°ç»„
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    # åˆå¹¶å·²æ’åºçš„æ•°ç»„
    return merge(left, right)

def merge(left, right):
    """åˆå¹¶ä¸¤ä¸ªå·²æ’åºæ•°ç»„"""
    result = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # æ·»åŠ å‰©ä½™å…ƒç´ 
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result

# æµ‹è¯•å½’å¹¶æ’åº
sorted_data5 = merge_sort(test_data)
print(f"   å½’å¹¶æ’åºç»“æœ: {sorted_data5}")

# æ’åºç®—æ³•æ€§èƒ½æ¯”è¾ƒ
print(f"\nğŸ“ 6. æ’åºç®—æ³•æ€§èƒ½æ¯”è¾ƒ:")
import time
import random

# ç”Ÿæˆæµ‹è¯•æ•°æ®
large_data = [random.randint(1, 1000) for _ in range(1000)]

# æµ‹è¯•å„ç®—æ³•æ€§èƒ½
algorithms = [
    ("å†’æ³¡æ’åº", bubble_sort),
    ("é€‰æ‹©æ’åº", selection_sort),
    ("æ’å…¥æ’åº", insertion_sort),
    ("å¿«é€Ÿæ’åº", quick_sort),
    ("å½’å¹¶æ’åº", merge_sort)
]

print(f"   æµ‹è¯•æ•°æ®å¤§å°: {len(large_data)}")
print(f"   ç®—æ³•æ€§èƒ½æ¯”è¾ƒ:")

for name, func in algorithms:
    start_time = time.time()
    result = func(large_data[:100])  # ä½¿ç”¨è¾ƒå°æ•°æ®æµ‹è¯•æ…¢ç®—æ³•
    end_time = time.time()
    
    print(f"     {name}: {end_time - start_time:.6f}ç§’")

# Pythonå†…ç½®æ’åº
start_time = time.time()
builtin_sorted = sorted(large_data)
end_time = time.time()
print(f"     Pythonå†…ç½®æ’åº: {end_time - start_time:.6f}ç§’")

print(f"\nâœ… åŸºç¡€æ’åºç®—æ³•å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å†’æ³¡æ’åºå®ç°")
print(f"   âœ“ ç†è§£é€‰æ‹©æ’åºåŸç†")
print(f"   âœ“ èƒ½åº”ç”¨å¿«é€Ÿå’Œå½’å¹¶æ’åº")
````

### æœç´¢ç®—æ³• [â­â­è¿›é˜¶]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šæœç´¢ç®—æ³•ç”¨äºåœ¨æ•°æ®é›†åˆä¸­æŸ¥æ‰¾ç‰¹å®šå…ƒç´ ï¼ŒåŒ…æ‹¬çº¿æ€§æœç´¢å’ŒäºŒåˆ†æœç´¢ç­‰ã€‚

**å­¦ä¹ è¦æ±‚**ï¼š
- æŒæ¡çº¿æ€§æœç´¢çš„å®ç°
- ç†è§£äºŒåˆ†æœç´¢çš„åŸç†
- èƒ½è¿›è¡Œæœç´¢ç®—æ³•çš„ä¼˜åŒ–

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°çº¿æ€§æœç´¢ç®—æ³•
- è¿›è¡ŒäºŒåˆ†æœç´¢åº”ç”¨
- åº”ç”¨æœç´¢è§£å†³å®é™…é—®é¢˜
- éªŒè¯ç‚¹ï¼šèƒ½æ­£ç¡®å®ç°æœç´¢ç®—æ³•

````python
print("ğŸ” æœç´¢ç®—æ³•:")
print("=" * 50)

# çº¿æ€§æœç´¢
print(f"ğŸ“ 1. çº¿æ€§æœç´¢:")
def linear_search(arr, target):
    """çº¿æ€§æœç´¢ç®—æ³•"""
    for i, value in enumerate(arr):
        if value == target:
            return i  # è¿”å›ç´¢å¼•
    return -1  # æœªæ‰¾åˆ°

# æµ‹è¯•çº¿æ€§æœç´¢
search_data = [10, 25, 3, 47, 15, 89, 23, 56, 78, 12]
target = 23
print(f"   æœç´¢æ•°æ®: {search_data}")
print(f"   ç›®æ ‡å€¼: {target}")

index = linear_search(search_data, target)
if index != -1:
    print(f"   çº¿æ€§æœç´¢ç»“æœ: æ‰¾åˆ°ï¼Œç´¢å¼•ä¸º {index}")
else:
    print(f"   çº¿æ€§æœç´¢ç»“æœ: æœªæ‰¾åˆ°")

# äºŒåˆ†æœç´¢
print(f"\nğŸ“ 2. äºŒåˆ†æœç´¢:")
def binary_search(arr, target):
    """äºŒåˆ†æœç´¢ç®—æ³•ï¼ˆè¦æ±‚æ•°ç»„å·²æ’åºï¼‰"""
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

# é€’å½’ç‰ˆæœ¬çš„äºŒåˆ†æœç´¢
def binary_search_recursive(arr, target, left=0, right=None):
    """é€’å½’äºŒåˆ†æœç´¢"""
    if right is None:
        right = len(arr) - 1
    
    if left > right:
        return -1
    
    mid = (left + right) // 2
    
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, target, mid + 1, right)
    else:
        return binary_search_recursive(arr, target, left, mid - 1)

# æµ‹è¯•äºŒåˆ†æœç´¢
sorted_data = sorted(search_data)
print(f"   æ’åºåæ•°æ®: {sorted_data}")

index2 = binary_search(sorted_data, target)
if index2 != -1:
    print(f"   äºŒåˆ†æœç´¢ç»“æœ: æ‰¾åˆ°ï¼Œç´¢å¼•ä¸º {index2}")
else:
    print(f"   äºŒåˆ†æœç´¢ç»“æœ: æœªæ‰¾åˆ°")

# é€’å½’äºŒåˆ†æœç´¢
index3 = binary_search_recursive(sorted_data, target)
print(f"   é€’å½’äºŒåˆ†æœç´¢ç»“æœ: æ‰¾åˆ°ï¼Œç´¢å¼•ä¸º {index3}")

# é«˜çº§æœç´¢ç®—æ³•
print(f"\nğŸ“ 3. é«˜çº§æœç´¢ç®—æ³•:")

# è·³è·ƒæœç´¢
def jump_search(arr, target):
    """è·³è·ƒæœç´¢ç®—æ³•"""
    import math
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    
    # è·³è·ƒæŸ¥æ‰¾ç›®æ ‡æ‰€åœ¨çš„å—
    while prev < n and arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1
    
    # åœ¨å—å†…çº¿æ€§æœç´¢
    while prev < min(step, n):
        if arr[prev] == target:
            return prev
        prev += 1
    
    return -1

# æ’å€¼æœç´¢
def interpolation_search(arr, target):
    """æ’å€¼æœç´¢ç®—æ³•"""
    left, right = 0, len(arr) - 1
    
    while left <= right and target >= arr[left] and target <= arr[right]:
        # ä¼°ç®—ç›®æ ‡ä½ç½®
        pos = left + ((target - arr[left]) * (right - left)) // (arr[right] - arr[left])
        
        if arr[pos] == target:
            return pos
        elif arr[pos] < target:
            left = pos + 1
        else:
            right = pos - 1
    
    return -1

# æµ‹è¯•é«˜çº§æœç´¢
print(f"   æµ‹è¯•è·³è·ƒæœç´¢:")
jump_result = jump_search(sorted_data, target)
print(f"     è·³è·ƒæœç´¢ç»“æœ: ç´¢å¼• {jump_result}")

print(f"\n   æµ‹è¯•æ’å€¼æœç´¢:")
interp_result = interpolation_search(sorted_data, target)
print(f"     æ’å€¼æœç´¢ç»“æœ: ç´¢å¼• {interp_result}")

# æœç´¢ç®—æ³•æ€§èƒ½æ¯”è¾ƒ
print(f"\nğŸ“ 4. æœç´¢ç®—æ³•æ€§èƒ½æ¯”è¾ƒ:")
import random

# ç”Ÿæˆå¤§æ•°æ®
large_sorted = sorted([random.randint(1, 10000) for _ in range(10000)])
search_target = large_sorted[5000]  # é€‰æ‹©ä¸­é—´çš„å…ƒç´ 

# æµ‹è¯•æ€§èƒ½
search_algorithms = [
    ("çº¿æ€§æœç´¢", linear_search),
    ("äºŒåˆ†æœç´¢", binary_search),
    ("é€’å½’äºŒåˆ†æœç´¢", binary_search_recursive),
    ("è·³è·ƒæœç´¢", jump_search),
    ("æ’å€¼æœç´¢", interpolation_search)
]

print(f"   æ•°æ®å¤§å°: {len(large_sorted)}")
print(f"   æœç´¢ç›®æ ‡: {search_target}")
print(f"   æ€§èƒ½æ¯”è¾ƒ:")

for name, func in search_algorithms:
    start_time = time.time()
    if name == "çº¿æ€§æœç´¢":
        result = func(large_sorted[:1000], search_target)  # çº¿æ€§æœç´¢ç”¨è¾ƒå°æ•°æ®
    else:
        result = func(large_sorted, search_target)
    end_time = time.time()
    
    print(f"     {name}: {end_time - start_time:.6f}ç§’, ç»“æœç´¢å¼•: {result}")

# å®é™…åº”ç”¨æ¡ˆä¾‹
print(f"\nğŸ“ 5. å®é™…åº”ç”¨æ¡ˆä¾‹:")

# å­¦ç”Ÿæˆç»©æœç´¢ç³»ç»Ÿ
class StudentSearchSystem:
    def __init__(self):
        self.students = []
        self.sorted_by_name = True
        self.sorted_by_score = True
    
    def add_student(self, name, score):
        """æ·»åŠ å­¦ç”Ÿ"""
        self.students.append((name, score))
        self.sorted_by_name = False
        self.sorted_by_score = False
    
    def search_by_name(self, name):
        """æŒ‰å§“åæœç´¢å­¦ç”Ÿ"""
        # çº¿æ€§æœç´¢ï¼ˆå› ä¸ºå§“åä¸æ˜¯æ’åºçš„ï¼‰
        results = []
        for i, (student_name, score) in enumerate(self.students):
            if name.lower() in student_name.lower():
                results.append((i, student_name, score))
        return results
    
    def search_by_score_range(self, min_score, max_score):
        """æŒ‰åˆ†æ•°èŒƒå›´æœç´¢"""
        if not self.sorted_by_score:
            self.students.sort(key=lambda x: x[1])
            self.sorted_by_score = True
        
        # ä½¿ç”¨äºŒåˆ†æœç´¢æ‰¾åˆ°èŒƒå›´
        scores = [score for _, score in self.students]
        
        # æ‰¾åˆ°æœ€å°åˆ†æ•°çš„èµ·å§‹ä½ç½®
        left = 0
        while left < len(scores) and scores[left] < min_score:
            left += 1
        
        # æ‰¾åˆ°æœ€å¤§åˆ†æ•°çš„ç»“æŸä½ç½®
        right = left
        while right < len(scores) and scores[right] <= max_score:
            right += 1
        
        return [(name, score) for name, score in self.students[left:right]]
    
    def get_top_students(self, n):
        """è·å–å‰nåå­¦ç”Ÿ"""
        if not self.sorted_by_score:
            self.students.sort(key=lambda x: x[1], reverse=True)
            self.sorted_by_score = True
        
        return self.students[:n]

# æµ‹è¯•å­¦ç”Ÿæœç´¢ç³»ç»Ÿ
search_system = StudentSearchSystem()

# æ·»åŠ å­¦ç”Ÿæ•°æ®
students_data = [
    ("å¼ ä¸‰", 85), ("æå››", 92), ("ç‹äº”", 78), ("èµµå…­", 88),
    ("é’±ä¸ƒ", 95), ("å­™å…«", 82), ("å‘¨ä¹", 90), ("å´å", 76),
    ("éƒ‘åä¸€", 89), ("ç‹åäºŒ", 84)
]

for name, score in students_data:
    search_system.add_student(name, score)

print(f"   å­¦ç”Ÿæ•°æ®å·²æ·»åŠ : {len(students_data)}åå­¦ç”Ÿ")

# æŒ‰å§“åæœç´¢
name_results = search_system.search_by_name("ç‹")
print(f"\n   å§“ååŒ…å«'ç‹'çš„å­¦ç”Ÿ:")
for idx, name, score in name_results:
    print(f"     {name}: {score}åˆ†")

# æŒ‰åˆ†æ•°èŒƒå›´æœç´¢
score_results = search_system.search_by_score_range(80, 90)
print(f"\n   åˆ†æ•°åœ¨80-90ä¹‹é—´çš„å­¦ç”Ÿ:")
for name, score in score_results:
    print(f"     {name}: {score}åˆ†")

# è·å–å‰3åå­¦ç”Ÿ
top_students = search_system.get_top_students(3)
print(f"\n   å‰3åå­¦ç”Ÿ:")
for i, (name, score) in enumerate(top_students, 1):
    print(f"     ç¬¬{i}å: {name}: {score}åˆ†")

print(f"\nâœ… æœç´¢ç®—æ³•å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡çº¿æ€§æœç´¢å®ç°")
print(f"   âœ“ ç†è§£äºŒåˆ†æœç´¢åŸç†")
print(f"   âœ“ èƒ½åº”ç”¨æœç´¢ç®—æ³•è§£å†³å®é™…é—®é¢˜")
````

````python
print("ğŸ”„ é€’å½’ç®—æ³•:")
print("=" * 50)

# åŸºç¡€é€’å½’ç¤ºä¾‹
print(f"ğŸ“ 1. åŸºç¡€é€’å½’ç¤ºä¾‹:")

# é˜¶ä¹˜è®¡ç®—
def factorial(n):
    """é€’å½’è®¡ç®—é˜¶ä¹˜"""
    if n <= 1:
        return 1
    return n * factorial(n - 1)

# æ–æ³¢é‚£å¥‘æ•°åˆ—
def fibonacci(n):
    """é€’å½’è®¡ç®—æ–æ³¢é‚£å¥‘æ•°åˆ—"""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

# å¹‚è¿ç®—
def power(x, n):
    """é€’å½’è®¡ç®—å¹‚è¿ç®—"""
    if n == 0:
        return 1
    elif n < 0:
        return 1 / power(x, -n)
    elif n % 2 == 0:
        half = power(x, n // 2)
        return half * half
    else:
        return x * power(x, n - 1)

# æµ‹è¯•åŸºç¡€é€’å½’
print(f"   é˜¶ä¹˜è®¡ç®—:")
for i in range(1, 8):
    print(f"     {i}! = {factorial(i)}")

print(f"\n   æ–æ³¢é‚£å¥‘æ•°åˆ—:")
fib_sequence = [fibonacci(i) for i in range(10)]
print(f"     {fib_sequence}")

print(f"\n   å¹‚è¿ç®—:")
print(f"     2^10 = {power(2, 10)}")
print(f"     3^5 = {power(3, 5)}")
print(f"     2^-3 = {power(2, -3):.3f}")

# é€’å½’ä¸è¿­ä»£å¯¹æ¯”
print(f"\nğŸ“ 2. é€’å½’ä¸è¿­ä»£å¯¹æ¯”:")

# è¿­ä»£ç‰ˆæœ¬çš„é˜¶ä¹˜
def factorial_iterative(n):
    """è¿­ä»£è®¡ç®—é˜¶ä¹˜"""
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# è¿­ä»£ç‰ˆæœ¬çš„æ–æ³¢é‚£å¥‘
def fibonacci_iterative(n):
    """è¿­ä»£è®¡ç®—æ–æ³¢é‚£å¥‘æ•°åˆ—"""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# æ€§èƒ½æ¯”è¾ƒ
test_n = 20
print(f"   æµ‹è¯•n={test_n}æ—¶çš„æ€§èƒ½:")

# é˜¶ä¹˜æ€§èƒ½æ¯”è¾ƒ
start_time = time.time()
fact_recursive = factorial(test_n)
recursive_time = time.time() - start_time

start_time = time.time()
fact_iterative = factorial_iterative(test_n)
iterative_time = time.time() - start_time

print(f"     é˜¶ä¹˜é€’å½’: {recursive_time:.6f}ç§’, ç»“æœ: {fact_recursive}")
print(f"     é˜¶ä¹˜è¿­ä»£: {iterative_time:.6f}ç§’, ç»“æœ: {fact_iterative}")

# æ–æ³¢é‚£å¥‘æ€§èƒ½æ¯”è¾ƒ
start_time = time.time()
fib_recursive = fibonacci(test_n)
fib_recursive_time = time.time() - start_time

start_time = time.time()
fib_iterative = fibonacci_iterative(test_n)
fib_iterative_time = time.time() - start_time

print(f"\n     æ–æ³¢é‚£å¥‘é€’å½’: {fib_recursive_time:.6f}ç§’, ç»“æœ: {fib_recursive}")
print(f"     æ–æ³¢é‚£å¥‘è¿­ä»£: {fib_iterative_time:.6f}ç§’, ç»“æœ: {fib_iterative}")

# é«˜çº§é€’å½’ç®—æ³•
print(f"\nğŸ“ 3. é«˜çº§é€’å½’ç®—æ³•:")

# æ±‰è¯ºå¡”é—®é¢˜
def hanoi(n, source, target, auxiliary):
    """æ±‰è¯ºå¡”é€’å½’è§£æ³•"""
    if n == 1:
        return [(source, target)]
    
    moves = []
    # å°†n-1ä¸ªç›˜å­ä»æºç§»åˆ°è¾…åŠ©æŸ±
    moves.extend(hanoi(n - 1, source, auxiliary, target))
    # å°†æœ€å¤§çš„ç›˜å­ä»æºç§»åˆ°ç›®æ ‡æŸ±
    moves.append((source, target))
    # å°†n-1ä¸ªç›˜å­ä»è¾…åŠ©æŸ±ç§»åˆ°ç›®æ ‡æŸ±
    moves.extend(hanoi(n - 1, auxiliary, target, source))
    
    return moves

# äºŒå‰æ ‘éå†
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def inorder_traversal(root):
    """ä¸­åºéå†äºŒå‰æ ‘"""
    if not root:
        return []
    
    result = []
    result.extend(inorder_traversal(root.left))
    result.append(root.val)
    result.extend(inorder_traversal(root.right))
    
    return result

def preorder_traversal(root):
    """å‰åºéå†äºŒå‰æ ‘"""
    if not root:
        return []
    
    result = [root.val]
    result.extend(preorder_traversal(root.left))
    result.extend(preorder_traversal(root.right))
    
    return result

def postorder_traversal(root):
    """ååºéå†äºŒå‰æ ‘"""
    if not root:
        return []
    
    result = []
    result.extend(postorder_traversal(root.left))
    result.extend(postorder_traversal(root.right))
    result.append(root.val)
    
    return result

# æµ‹è¯•æ±‰è¯ºå¡”
print(f"   æ±‰è¯ºå¡”é—®é¢˜(3ä¸ªç›˜å­):")
hanoi_moves = hanoi(3, 'A', 'C', 'B')
print(f"     ç§»åŠ¨æ­¥éª¤: {len(hanoi_moves)}æ­¥")
for i, (source, target) in enumerate(hanoi_moves[:10], 1):
    print(f"     ç¬¬{i}æ­¥: {source} -> {target}")
if len(hanoi_moves) > 10:
    print(f"     ... è¿˜æœ‰{len(hanoi_moves) - 10}æ­¥")

# æµ‹è¯•äºŒå‰æ ‘éå†
print(f"\n   äºŒå‰æ ‘éå†:")
# æ„å»ºç¤ºä¾‹äºŒå‰æ ‘
#       4
#      / \
#     2   6
#    / \ / \
#   1  3 5  7
tree_root = TreeNode(4)
tree_root.left = TreeNode(2)
tree_root.right = TreeNode(6)
tree_root.left.left = TreeNode(1)
tree_root.left.right = TreeNode(3)
tree_root.right.left = TreeNode(5)
tree_root.right.right = TreeNode(7)

inorder = inorder_traversal(tree_root)
preorder = preorder_traversal(tree_root)
postorder = postorder_traversal(tree_root)

print(f"     ä¸­åºéå†: {inorder}")
print(f"     å‰åºéå†: {preorder}")
print(f"     ååºéå†: {postorder}")

# é€’å½’ä¼˜åŒ–æŠ€æœ¯
print(f"\nğŸ“ 4. é€’å½’ä¼˜åŒ–æŠ€æœ¯:")

# è®°å¿†åŒ–é€’å½’ï¼ˆåŠ¨æ€è§„åˆ’ï¼‰
def fibonacci_memo(n, memo={}):
    """è®°å¿†åŒ–é€’å½’æ–æ³¢é‚£å¥‘"""
    if n in memo:
        return memo[n]
    
    if n <= 0:
        memo[0] = 0
        return 0
    elif n == 1:
        memo[1] = 1
        return 1
    
    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]

# å°¾é€’å½’ä¼˜åŒ–ï¼ˆPythonä¸æ”¯æŒçœŸæ­£çš„å°¾é€’å½’ä¼˜åŒ–ï¼Œä½†å¯ä»¥æ¨¡æ‹Ÿï¼‰
def factorial_tail(n, accumulator=1):
    """å°¾é€’å½’é˜¶ä¹˜"""
    if n <= 1:
        return accumulator
    return factorial_tail(n - 1, n * accumulator)

# æ€§èƒ½æ¯”è¾ƒ
test_fib_n = 35
print(f"   æ–æ³¢é‚£å¥‘æ€§èƒ½æ¯”è¾ƒ(n={test_fib_n}):")

# æ™®é€šé€’å½’
start_time = time.time()
fib_normal = fibonacci(test_fib_n)
normal_time = time.time() - start_time

# è®°å¿†åŒ–é€’å½’
start_time = time.time()
fib_memo = fibonacci_memo(test_fib_n)
memo_time = time.time() - start_time

print(f"     æ™®é€šé€’å½’: {normal_time:.6f}ç§’, ç»“æœ: {fib_normal}")
print(f"     è®°å¿†åŒ–é€’å½’: {memo_time:.6f}ç§’, ç»“æœ: {fib_memo}")
print(f"     æ€§èƒ½æå‡: {normal_time/memo_time:.1f}å€")

# å®é™…åº”ç”¨æ¡ˆä¾‹
print(f"\nğŸ“ 5. å®é™…åº”ç”¨æ¡ˆä¾‹:")

# ç›®å½•ç»“æ„éå†
class FileSystemNode:
    def __init__(self, name, is_file=False):
        self.name = name
        self.is_file = is_file
        self.children = []
    
    def add_child(self, child):
        self.children.append(child)
    
    def find_files(self, extension=None):
        """é€’å½’æŸ¥æ‰¾æ–‡ä»¶"""
        files = []
        
        if self.is_file:
            if extension is None or self.name.endswith(extension):
                files.append(self.name)
        else:
            for child in self.children:
                files.extend(child.find_files(extension))
        
        return files
    
    def get_size(self):
        """é€’å½’è®¡ç®—ç›®å½•å¤§å°"""
        if self.is_file:
            return 1  # å‡è®¾æ¯ä¸ªæ–‡ä»¶å¤§å°ä¸º1
        
        total_size = 0
        for child in self.children:
            total_size += child.get_size()
        
        return total_size

# æ„å»ºç¤ºä¾‹æ–‡ä»¶ç³»ç»Ÿ
root = FileSystemNode("root")
docs = FileSystemNode("documents")
code = FileSystemNode("code")
images = FileSystemNode("images")

root.add_child(docs)
root.add_child(code)
root.add_child(images)

# æ·»åŠ æ–‡ä»¶
docs.add_child(FileSystemNode("readme.txt", True))
docs.add_child(FileSystemNode("manual.pdf", True))
code.add_child(FileSystemNode("main.py", True))
code.add_child(FileSystemNode("utils.py", True))
code.add_child(FileSystemNode("test.py", True))
images.add_child(FileSystemNode("logo.png", True))
images.add_child(FileSystemNode("banner.jpg", True))

print(f"   æ–‡ä»¶ç³»ç»Ÿç»“æ„:")
all_files = root.find_files()
print(f"     æ‰€æœ‰æ–‡ä»¶: {all_files}")

py_files = root.find_files(".py")
print(f"\n     Pythonæ–‡ä»¶: {py_files}")

total_size = root.get_size()
print(f"\n   æ€»æ–‡ä»¶æ•°: {total_size}")

for child in root.children:
    child_size = child.get_size()
    print(f"     {child.name}: {child_size}ä¸ªæ–‡ä»¶")

print(f"\nâœ… é€’å½’ç®—æ³•å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡é€’å½’åŸºæœ¬æ¦‚å¿µ")
print(f"   âœ“ ç†è§£é€’å½’é—®é¢˜è§£å†³æ€è·¯")
print(f"   âœ“ èƒ½è®¾è®¡å’Œå®ç°é€’å½’ç®—æ³•")
````

### ç»¼åˆç®—æ³•åº”ç”¨ [â­â­è¿›é˜¶]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šé€šè¿‡ç»¼åˆåº”ç”¨æ¡ˆä¾‹å±•ç¤ºç®—æ³•åœ¨å®é™…é—®é¢˜ä¸­çš„ç»„åˆä½¿ç”¨å’Œä¼˜åŒ–ç­–ç•¥ã€‚

**å­¦ä¹ è¦æ±‚**ï¼š
- æŒæ¡ç®—æ³•çš„ç»¼åˆåº”ç”¨
- ç†è§£ç®—æ³•é€‰æ‹©å’Œä¼˜åŒ–
- èƒ½è§£å†³å¤æ‚ç®—æ³•é—®é¢˜

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°ç»¼åˆç®—æ³•è§£å†³æ–¹æ¡ˆ
- è¿›è¡Œç®—æ³•æ€§èƒ½ä¼˜åŒ–
- åº”ç”¨ç®—æ³•è§£å†³å®é™…é—®é¢˜
- éªŒè¯ç‚¹ï¼šèƒ½ç‹¬ç«‹è§£å†³å¤æ‚ç®—æ³•é—®é¢˜

````python
print("ğŸ¯ ç»¼åˆç®—æ³•åº”ç”¨:")
print("=" * 50)

# æ¡ˆä¾‹1ï¼šæ•°æ®åˆ†æç®¡é“
print(f"ğŸ“ 1. æ•°æ®åˆ†æç®¡é“:")

class DataAnalysisPipeline:
    def __init__(self):
        self.data = []
        self.processed_data = []
    
    def load_data(self, data):
        """åŠ è½½æ•°æ®"""
        self.data = data.copy()
        print(f"   åŠ è½½æ•°æ®: {len(self.data)}æ¡è®°å½•")
    
    def clean_data(self):
        """æ•°æ®æ¸…æ´—"""
        # ç§»é™¤æ— æ•ˆæ•°æ®
        cleaned = [x for x in self.data if isinstance(x, (int, float)) and x >= 0]
        print(f"   æ¸…æ´—åæ•°æ®: {len(cleaned)}æ¡è®°å½•")
        
        # ç§»é™¤å¼‚å¸¸å€¼ï¼ˆä½¿ç”¨IQRæ–¹æ³•ï¼‰
        if cleaned:
            sorted_data = sorted(cleaned)
            n = len(sorted_data)
            q1 = sorted_data[n // 4]
            q3 = sorted_data[3 * n // 4]
            iqr = q3 - q1
            
            lower_bound = q1 - 1.5 * iqr
            upper_bound = q3 + 1.5 * iqr
            
            filtered = [x for x in cleaned if lower_bound <= x <= upper_bound]
            print(f"   ç§»é™¤å¼‚å¸¸å€¼å: {len(filtered)}æ¡è®°å½•")
            
            self.processed_data = filtered
        else:
            self.processed_data = cleaned
    
    def analyze_data(self):
        """æ•°æ®åˆ†æ"""
        if not self.processed_data:
            return None
        
        # åŸºç¡€ç»Ÿè®¡
        n = len(self.processed_data)
        total = sum(self.processed_data)
        mean = total / n
        
        # æ’åºæ•°æ®ç”¨äºè®¡ç®—ä¸­ä½æ•°å’Œåˆ†ä½æ•°
        sorted_data = sorted(self.processed_data)
        median = sorted_data[n // 2] if n % 2 == 1 else (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
        
        # è®¡ç®—æ ‡å‡†å·®
        variance = sum((x - mean) ** 2 for x in self.processed_data) / n
        std_dev = variance ** 0.5
        
        # æ•°æ®åˆ†å¸ƒåˆ†æ
        min_val = min(self.processed_data)
        max_val = max(self.processed_data)
        
        # å°†æ•°æ®åˆ†ä¸ºå‡ ä¸ªåŒºé—´
        range_size = (max_val - min_val) / 5
        bins = []
        for i in range(5):
            lower = min_val + i * range_size
            upper = min_val + (i + 1) * range_size
            count = sum(1 for x in self.processed_data if lower <= x < upper)
            bins.append((lower, upper, count))
        
        return {
            'count': n,
            'mean': mean,
            'median': median,
            'std_dev': std_dev,
            'min': min_val,
            'max': max_val,
            'distribution': bins
        }
    
    def find_patterns(self):
        """å¯»æ‰¾æ•°æ®æ¨¡å¼"""
        if not self.processed_data:
            return None
        
        # å¯»æ‰¾é‡å¤å€¼
        value_counts = {}
        for value in self.processed_data:
            value_counts[value] = value_counts.get(value, 0) + 1
        
        # æ‰¾å‡ºå‡ºç°é¢‘ç‡æœ€é«˜çš„å€¼
        most_common = max(value_counts.items(), key=lambda x: x[1])
        
        # å¯»æ‰¾è¶‹åŠ¿ï¼ˆç®€å•çº¿æ€§å›å½’ï¼‰
        n = len(self.processed_data)
        x_mean = (n - 1) / 2
        y_mean = sum(self.processed_data) / n
        
        numerator = sum((i - x_mean) * (self.processed_data[i] - y_mean) for i in range(n))
        denominator = sum((i - x_mean) ** 2 for i in range(n))
        
        slope = numerator / denominator if denominator != 0 else 0
        
        return {
            'most_common': most_common,
            'trend_slope': slope,
            'unique_values': len(value_counts)
        }

# æµ‹è¯•æ•°æ®åˆ†æç®¡é“
pipeline = DataAnalysisPipeline()

# ç”Ÿæˆæµ‹è¯•æ•°æ®
import random
test_data = [random.randint(10, 100) for _ in range(50)]
test_data.extend([150, 200, 5, -10])  # æ·»åŠ ä¸€äº›å¼‚å¸¸å€¼
test_data.extend([45, 45, 45, 60, 60])  # æ·»åŠ é‡å¤å€¼

pipeline.load_data(test_data)
pipeline.clean_data()

analysis = pipeline.analyze_data()
if analysis:
    print(f"\n   æ•°æ®åˆ†æç»“æœ:")
    print(f"     æ•°æ®é‡: {analysis['count']}")
    print(f"     å¹³å‡å€¼: {analysis['mean']:.2f}")
    print(f"     ä¸­ä½æ•°: {analysis['median']:.2f}")
    print(f"     æ ‡å‡†å·®: {analysis['std_dev']:.2f}")
    print(f"     èŒƒå›´: {analysis['min']:.2f} - {analysis['max']:.2f}")
    
    print(f"\n     æ•°æ®åˆ†å¸ƒ:")
    for i, (lower, upper, count) in enumerate(analysis['distribution']):
        print(f"       åŒºé—´{i+1}: {lower:.1f}-{upper:.1f}: {count}ä¸ªæ•°æ®")

patterns = pipeline.find_patterns()
if patterns:
    print(f"\n   æ•°æ®æ¨¡å¼:")
    print(f"     æœ€å¸¸è§å€¼: {patterns['most_common'][0]} (å‡ºç°{patterns['most_common'][1]}æ¬¡)")
    print(f"     è¶‹åŠ¿æ–œç‡: {patterns['trend_slope']:.4f}")
    print(f"     å”¯ä¸€å€¼æ•°é‡: {patterns['unique_values']}")

# æ¡ˆä¾‹2ï¼šæ™ºèƒ½æ¨èç³»ç»Ÿ
print(f"\nğŸ“ 2. æ™ºèƒ½æ¨èç³»ç»Ÿ:")

class RecommendationSystem:
    def __init__(self):
        self.users = {}
        self.items = {}
        self.ratings = {}
    
    def add_user(self, user_id, preferences=None):
        """æ·»åŠ ç”¨æˆ·"""
        self.users[user_id] = preferences or {}
    
    def add_item(self, item_id, categories=None):
        """æ·»åŠ ç‰©å“"""
        self.items[item_id] = categories or []
    
    def add_rating(self, user_id, item_id, rating):
        """æ·»åŠ è¯„åˆ†"""
        if user_id not in self.ratings:
            self.ratings[user_id] = {}
        self.ratings[user_id][item_id] = rating
    
    def calculate_similarity(self, user1, user2):
        """è®¡ç®—ç”¨æˆ·ç›¸ä¼¼åº¦ï¼ˆåŸºäºå…±åŒè¯„åˆ†ï¼‰"""
        if user1 not in self.ratings or user2 not in self.ratings:
            return 0
        
        ratings1 = self.ratings[user1]
        ratings2 = self.ratings[user2]
        
        # æ‰¾åˆ°å…±åŒè¯„åˆ†çš„ç‰©å“
        common_items = set(ratings1.keys()) & set(ratings2.keys())
        
        if not common_items:
            return 0
        
        # è®¡ç®—çš®å°”é€Šç›¸å…³ç³»æ•°
        sum1 = sum(ratings1[item] for item in common_items)
        sum2 = sum(ratings2[item] for item in common_items)
        
        sum1_sq = sum(ratings1[item] ** 2 for item in common_items)
        sum2_sq = sum(ratings2[item] ** 2 for item in common_items)
        
        sum_products = sum(ratings1[item] * ratings2[item] for item in common_items)
        
        n = len(common_items)
        numerator = sum_products - (sum1 * sum2 / n)
        denominator = ((sum1_sq - sum1 ** 2 / n) * (sum2_sq - sum2 ** 2 / n)) ** 0.5
        
        return numerator / denominator if denominator != 0 else 0
    
    def find_similar_users(self, user_id, k=5):
        """æ‰¾åˆ°ç›¸ä¼¼ç”¨æˆ·"""
        similarities = []
        
        for other_user in self.ratings:
            if other_user != user_id:
                similarity = self.calculate_similarity(user_id, other_user)
                similarities.append((other_user, similarity))
        
        # æŒ‰ç›¸ä¼¼åº¦æ’åº
        similarities.sort(key=lambda x: x[1], reverse=True)
        return similarities[:k]
    
    def recommend_items(self, user_id, k=5):
        """ä¸ºç”¨æˆ·æ¨èç‰©å“"""
        if user_id not in self.ratings:
            return []
        
        # æ‰¾åˆ°ç›¸ä¼¼ç”¨æˆ·
        similar_users = self.find_similar_users(user_id, k=3)
        
        # æ”¶é›†ç›¸ä¼¼ç”¨æˆ·è¯„åˆ†ä½†ç›®æ ‡ç”¨æˆ·æœªè¯„åˆ†çš„ç‰©å“
        user_ratings = self.ratings[user_id]
        recommendations = {}
        
        for similar_user, similarity in similar_users:
            if similarity > 0:
                for item, rating in self.ratings[similar_user].items():
                    if item not in user_ratings:
                        if item not in recommendations:
                            recommendations[item] = 0
                        recommendations[item] += similarity * rating
        
        # æŒ‰æ¨èåˆ†æ•°æ’åº
        sorted_recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)
        return sorted_recommendations[:k]

# æµ‹è¯•æ¨èç³»ç»Ÿ
rec_system = RecommendationSystem()

# æ·»åŠ ç”¨æˆ·å’Œç‰©å“
users = ["ç”¨æˆ·1", "ç”¨æˆ·2", "ç”¨æˆ·3", "ç”¨æˆ·4", "ç”¨æˆ·5"]
items = ["ç‰©å“A", "ç‰©å“B", "ç‰©å“C", "ç‰©å“D", "ç‰©å“E", "ç‰©å“F"]

for user in users:
    rec_system.add_user(user)

for item in items:
    rec_system.add_item(item)

# æ·»åŠ è¯„åˆ†ï¼ˆæ¨¡æ‹Ÿç”¨æˆ·åå¥½ï¼‰
ratings_data = {
    "ç”¨æˆ·1": {"ç‰©å“A": 5, "ç‰©å“B": 3, "ç‰©å“C": 4, "ç‰©å“D": 2},
    "ç”¨æˆ·2": {"ç‰©å“A": 4, "ç‰©å“B": 2, "ç‰©å“C": 5, "ç‰©å“E": 3},
    "ç”¨æˆ·3": {"ç‰©å“B": 5, "ç‰©å“C": 3, "ç‰©å“D": 4, "ç‰©å“F": 5},
    "ç”¨æˆ·4": {"ç‰©å“A": 3, "ç‰©å“D": 5, "ç‰©å“E": 4, "ç‰©å“F": 2},
    "ç”¨æˆ·5": {"ç‰©å“B": 4, "ç‰©å“C": 5, "ç‰©å“E": 5, "ç‰©å“F": 4}
}

for user, user_ratings in ratings_data.items():
    for item, rating in user_ratings.items():
        rec_system.add_rating(user, item, rating)

print(f"   æ¨èç³»ç»Ÿå·²åˆå§‹åŒ–: {len(users)}ç”¨æˆ·, {len(items)}ç‰©å“")

# ä¸ºç”¨æˆ·1æ¨èç‰©å“
recommendations = rec_system.recommend_items("ç”¨æˆ·1")
print(f"\n   ä¸ºç”¨æˆ·1æ¨èçš„ç‰©å“:")
for item, score in recommendations:
    print(f"     {item}: æ¨èåˆ†æ•° {score:.2f}")

# æ‰¾åˆ°ç›¸ä¼¼ç”¨æˆ·
similar_users = rec_system.find_similar_users("ç”¨æˆ·1")
print(f"\n   ä¸ç”¨æˆ·1ç›¸ä¼¼çš„ç”¨æˆ·:")
for user, similarity in similar_users:
    print(f"     {user}: ç›¸ä¼¼åº¦ {similarity:.3f}")

# æ¡ˆä¾‹3ï¼šè·¯å¾„è§„åˆ’ç®—æ³•
print(f"\nğŸ“ 3. è·¯å¾„è§„åˆ’ç®—æ³•:")

class PathFinder:
    def __init__(self, graph=None):
        self.graph = graph or {}
    
    def add_edge(self, from_node, to_node, weight):
        """æ·»åŠ è¾¹"""
        if from_node not in self.graph:
            self.graph[from_node] = {}
        if to_node not in self.graph:
            self.graph[to_node] = {}
        
        self.graph[from_node][to_node] = weight
        self.graph[to_node][from_node] = weight  # æ— å‘å›¾
    
    def dijkstra(self, start, end):
        """Dijkstraæœ€çŸ­è·¯å¾„ç®—æ³•"""
        import heapq
        
        # åˆå§‹åŒ–è·ç¦»å­—å…¸
        distances = {node: float('infinity') for node in self.graph}
        distances[start] = 0
        
        # ä¼˜å…ˆé˜Ÿåˆ—ï¼š(è·ç¦», èŠ‚ç‚¹)
        pq = [(0, start)]
        
        # è®°å½•è·¯å¾„
        previous = {}
        
        while pq:
            current_distance, current_node = heapq.heappop(pq)
            
            # å¦‚æœæ‰¾åˆ°æ›´çŸ­çš„è·¯å¾„ï¼Œè·³è¿‡
            if current_distance > distances[current_node]:
                continue
            
            # åˆ°è¾¾ç›®æ ‡èŠ‚ç‚¹
            if current_node == end:
                break
            
            # æ£€æŸ¥é‚»å±…
            for neighbor, weight in self.graph[current_node].items():
                distance = current_distance + weight
                
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    previous[neighbor] = current_node
                    heapq.heappush(pq, (distance, neighbor))
        
        # æ„å»ºè·¯å¾„
        path = []
        current = end
        while current in previous:
            path.append(current)
            current = previous[current]
        path.append(start)
        path.reverse()
        
        return path, distances[end]
    
    def find_all_paths(self, start, end, max_depth=10):
        """æŸ¥æ‰¾æ‰€æœ‰è·¯å¾„ï¼ˆé€’å½’ï¼‰"""
        def dfs(current, target, visited, path, all_paths, depth):
            if depth > max_depth:
                return
            
            visited.add(current)
            path.append(current)
            
            if current == target:
                all_paths.append(path.copy())
            else:
                for neighbor in self.graph[current]:
                    if neighbor not in visited:
                        dfs(neighbor, target, visited, path, all_paths, depth + 1)
            
            path.pop()
            visited.remove(current)
        
        all_paths = []
        dfs(start, end, set(), [], all_paths, 0)
        return all_paths

# æµ‹è¯•è·¯å¾„è§„åˆ’
path_finder = PathFinder()

# æ„å»ºå›¾
edges = [
    ("A", "B", 2), ("A", "C", 4), ("B", "C", 1),
    ("B", "D", 7), ("C", "D", 3), ("C", "E", 5),
    ("D", "E", 1), ("D", "F", 2), ("E", "F", 8)
]

for from_node, to_node, weight in edges:
    path_finder.add_edge(from_node, to_node, weight)

print(f"   å›¾å·²æ„å»º: {len(path_finder.graph)}ä¸ªèŠ‚ç‚¹")

# æŸ¥æ‰¾æœ€çŸ­è·¯å¾„
start_node = "A"
end_node = "F"

shortest_path, distance = path_finder.dijkstra(start_node, end_node)
print(f"\n   ä»{start_node}åˆ°{end_node}çš„æœ€çŸ­è·¯å¾„:")
print(f"     è·¯å¾„: {' -> '.join(shortest_path)}")
print(f"     æ€»è·ç¦»: {distance}")

# æŸ¥æ‰¾æ‰€æœ‰è·¯å¾„
all_paths = path_finder.find_all_paths(start_node, end_node)
print(f"\n   ä»{start_node}åˆ°{end_node}çš„æ‰€æœ‰è·¯å¾„:")
for i, path in enumerate(all_paths, 1):
    print(f"     è·¯å¾„{i}: {' -> '.join(path)}")

print(f"\nâœ… ç»¼åˆç®—æ³•åº”ç”¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡ç®—æ³•ç»¼åˆåº”ç”¨")
print(f"   âœ“ ç†è§£ç®—æ³•é€‰æ‹©å’Œä¼˜åŒ–")
print(f"   âœ“ èƒ½è§£å†³å¤æ‚ç®—æ³•é—®é¢˜")
````
