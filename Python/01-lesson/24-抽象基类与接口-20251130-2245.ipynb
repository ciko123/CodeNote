````python
print("ğŸ”§ æ¥å£è®¾è®¡æ¨¡å¼:")
print("=" * 50)

# 1. ç­–ç•¥æ¨¡å¼ (Strategy Pattern)
print(f"ğŸ“ 1. ç­–ç•¥æ¨¡å¼ (Strategy Pattern):")

class PaymentStrategy(ABC):
    """æ”¯ä»˜ç­–ç•¥æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def pay(self, amount):
        """æ”¯ä»˜"""
        pass
    
    @abstractmethod
    def get_payment_info(self):
        """è·å–æ”¯ä»˜ä¿¡æ¯"""
        pass

class CreditCardPayment(PaymentStrategy):
    """ä¿¡ç”¨å¡æ”¯ä»˜ç­–ç•¥"""
    
    def __init__(self, card_number, card_holder):
        self.card_number = card_number
        self.card_holder = card_holder
    
    def pay(self, amount):
        return f"ä½¿ç”¨ä¿¡ç”¨å¡{self.card_number}æ”¯ä»˜{amount}å…ƒ"
    
    def get_payment_info(self):
        return f"ä¿¡ç”¨å¡æ”¯ä»˜ - æŒå¡äºº: {self.card_holder}"

class AlipayPayment(PaymentStrategy):
    """æ”¯ä»˜å®æ”¯ä»˜ç­–ç•¥"""
    
    def __init__(self, account_id):
        self.account_id = account_id
    
    def pay(self, amount):
        return f"ä½¿ç”¨æ”¯ä»˜å®{self.account_id}æ”¯ä»˜{amount}å…ƒ"
    
    def get_payment_info(self):
        return f"æ”¯ä»˜å®æ”¯ä»˜ - è´¦æˆ·: {self.account_id}"

class WeChatPayment(PaymentStrategy):
    """å¾®ä¿¡æ”¯ä»˜ç­–ç•¥"""
    
    def __init__(self, openid):
        self.openid = openid
    
    def pay(self, amount):
        return f"ä½¿ç”¨å¾®ä¿¡{self.openid}æ”¯ä»˜{amount}å…ƒ"
    
    def get_payment_info(self):
        return f"å¾®ä¿¡æ”¯ä»˜ - ç”¨æˆ·: {self.openid}"

class PaymentContext:
    """æ”¯ä»˜ä¸Šä¸‹æ–‡"""
    
    def __init__(self, strategy: PaymentStrategy):
        self._strategy = strategy
    
    def set_strategy(self, strategy: PaymentStrategy):
        """è®¾ç½®æ”¯ä»˜ç­–ç•¥"""
        self._strategy = strategy
    
    def execute_payment(self, amount):
        """æ‰§è¡Œæ”¯ä»˜"""
        if not self._strategy:
            raise ValueError("æœªè®¾ç½®æ”¯ä»˜ç­–ç•¥")
        return self._strategy.pay(amount)
    
    def get_payment_method(self):
        """è·å–æ”¯ä»˜æ–¹å¼ä¿¡æ¯"""
        if not self._strategy:
            return "æœªè®¾ç½®æ”¯ä»˜æ–¹å¼"
        return self._strategy.get_payment_info()

# æµ‹è¯•ç­–ç•¥æ¨¡å¼
print(f"   ç­–ç•¥æ¨¡å¼æµ‹è¯•:")

# åˆ›å»ºæ”¯ä»˜ç­–ç•¥
credit_card = CreditCardPayment("1234567890123456", "å¼ ä¸‰")
alipay = AlipayPayment("zhangsan@example.com")
wechat = WeChatPayment("wx_zhangsan")

# åˆ›å»ºæ”¯ä»˜ä¸Šä¸‹æ–‡
payment_context = PaymentContext(credit_card)

print(f"\n     æ”¯ä»˜æ–¹å¼: {payment_context.get_payment_method()}")
print(f"       {payment_context.execute_payment(100)}")

# åˆ‡æ¢æ”¯ä»˜ç­–ç•¥
payment_context.set_strategy(alipay)
print(f"\n     æ”¯ä»˜æ–¹å¼: {payment_context.get_payment_method()}")
print(f"       {payment_context.execute_payment(200)}")

payment_context.set_strategy(wechat)
print(f"\n     æ”¯ä»˜æ–¹å¼: {payment_context.get_payment_method()}")
print(f"       {payment_context.execute_payment(300)}")

# 2. è§‚å¯Ÿè€…æ¨¡å¼ (Observer Pattern)
print(f"\nğŸ“ 2. è§‚å¯Ÿè€…æ¨¡å¼ (Observer Pattern):")

class Observer(ABC):
    """è§‚å¯Ÿè€…æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def update(self, message):
        """æ›´æ–°é€šçŸ¥"""
        pass
    
    @abstractmethod
    def get_observer_id(self):
        """è·å–è§‚å¯Ÿè€…ID"""
        pass

class Subject(ABC):
    """ä¸»é¢˜æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def attach(self, observer: Observer):
        """æ·»åŠ è§‚å¯Ÿè€…"""
        pass
    
    @abstractmethod
    def detach(self, observer: Observer):
        """ç§»é™¤è§‚å¯Ÿè€…"""
        pass
    
    @abstractmethod
    def notify(self, message):
        """é€šçŸ¥è§‚å¯Ÿè€…"""
        pass

class EmailObserver(Observer):
    """é‚®ä»¶è§‚å¯Ÿè€…"""
    
    def __init__(self, email):
        self.email = email
        self.notifications = []
    
    def update(self, message):
        notification = f"é‚®ä»¶å‘é€åˆ°{self.email}: {message}"
        self.notifications.append(notification)
        print(f"       {notification}")
    
    def get_observer_id(self):
        return self.email
    
    def get_notifications(self):
        return self.notifications.copy()

class SMSObserver(Observer):
    """çŸ­ä¿¡è§‚å¯Ÿè€…"""
    
    def __init__(self, phone):
        self.phone = phone
        self.notifications = []
    
    def update(self, message):
        notification = f"çŸ­ä¿¡å‘é€åˆ°{self.phone}: {message}"
        self.notifications.append(notification)
        print(f"       {notification}")
    
    def get_observer_id(self):
        return self.phone
    
    def get_notifications(self):
        return self.notifications.copy()

class PushObserver(Observer):
    """æ¨é€è§‚å¯Ÿè€…"""
    
    def __init__(self, device_id):
        self.device_id = device_id
        self.notifications = []
    
    def update(self, message):
        notification = f"æ¨é€åˆ°è®¾å¤‡{self.device_id}: {message}"
        self.notifications.append(notification)
        print(f"       {notification}")
    
    def get_observer_id(self):
        return self.device_id
    
    def get_notifications(self):
        return self.notifications.copy()

class NewsAgency(Subject):
    """æ–°é—»æœºæ„ - ä¸»é¢˜å®ç°"""
    
    def __init__(self, name):
        self.name = name
        self._observers = []
    
    def attach(self, observer: Observer):
        if observer not in self._observers:
            self._observers.append(observer)
            print(f"       {observer.get_observer_id()}å·²è®¢é˜…{self.name}")
        else:
            print(f"       {observer.get_observer_id()}å·²è®¢é˜…{self.name}")
    
    def detach(self, observer: Observer):
        if observer in self._observers:
            self._observers.remove(observer)
            print(f"       {observer.get_observer_id()}å·²å–æ¶ˆè®¢é˜…{self.name}")
        else:
            print(f"       {observer.get_observer_id()}æœªè®¢é˜…{self.name}")
    
    def notify(self, message):
        print(f"\n       {self.name}å‘å¸ƒæ–°é—»: {message}")
        for observer in self._observers:
            observer.update(message)
    
    def publish_news(self, headline):
        """å‘å¸ƒæ–°é—»"""
        self.notify(headline)
    
    def get_subscribers(self):
        """è·å–è®¢é˜…è€…åˆ—è¡¨"""
        return [observer.get_observer_id() for observer in self._observers]

# æµ‹è¯•è§‚å¯Ÿè€…æ¨¡å¼
print(f"\n   è§‚å¯Ÿè€…æ¨¡å¼æµ‹è¯•:")

# åˆ›å»ºè§‚å¯Ÿè€…
email_obs = EmailObserver("user@example.com")
sms_obs = SMSObserver("13800138000")
push_obs = PushObserver("device123")

# åˆ›å»ºä¸»é¢˜
news_agency = NewsAgency("æ–°åæ–°é—»")

# æ·»åŠ è§‚å¯Ÿè€…
news_agency.attach(email_obs)
news_agency.attach(sms_obs)
news_agency.attach(push_obs)

print(f"\n     å½“å‰è®¢é˜…è€…: {news_agency.get_subscribers()}")

# å‘å¸ƒæ–°é—»
news_agency.publish_news("ä»Šæ—¥å¤´æ¡ï¼šPythonå­¦ä¹ æŒ‡å—å‘å¸ƒ")
news_agency.publish_news("ç§‘æŠ€æ–°é—»ï¼šAIæŠ€æœ¯æ–°çªç ´")

# ç§»é™¤è§‚å¯Ÿè€…
news_agency.detach(sms_obs)
print(f"\n     ç§»é™¤åè®¢é˜…è€…: {news_agency.get_subscribers()}")

# å†æ¬¡å‘å¸ƒæ–°é—»
news_agency.publish_news("ä½“è‚²æ–°é—»ï¼šä¸–ç•Œæ¯ç²¾å½©èµ›äº‹")

# 3. å‘½ä»¤æ¨¡å¼ (Command Pattern)
print(f"\nğŸ“ 3. å‘½ä»¤æ¨¡å¼ (Command Pattern):")

class Command(ABC):
    """å‘½ä»¤æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def execute(self):
        """æ‰§è¡Œå‘½ä»¤"""
        pass
    
    @abstractmethod
    def undo(self):
        """æ’¤é”€å‘½ä»¤"""
        pass
    
    @abstractmethod
    def get_description(self):
        """è·å–å‘½ä»¤æè¿°"""
        pass

class Light:
    """ç”µç¯ç±» - æ¥æ”¶è€…"""
    
    def __init__(self, location):
        self.location = location
        self.is_on = False
    
    def turn_on(self):
        self.is_on = True
        print(f"       {self.location}çš„ç¯å·²æ‰“å¼€")
    
    def turn_off(self):
        self.is_on = False
        print(f"       {self.location}çš„ç¯å·²å…³é—­")
    
    def get_status(self):
        return f"{self.location}çš„ç¯{'å¼€' if self.is_on else 'å…³'}"

class LightOnCommand(Command):
    """å¼€ç¯å‘½ä»¤"""
    
    def __init__(self, light: Light):
        self.light = light
    
    def execute(self):
        self.light.turn_on()
    
    def undo(self):
        self.light.turn_off()
    
    def get_description(self):
        return f"æ‰“å¼€{self.light.location}çš„ç¯"

class LightOffCommand(Command):
    """å…³ç¯å‘½ä»¤"""
    
    def __init__(self, light: Light):
        self.light = light
    
    def execute(self):
        self.light.turn_off()
    
    def undo(self):
        self.light.turn_on()
    
    def get_description(self):
        return f"å…³é—­{self.light.location}çš„ç¯"

class MacroCommand(Command):
    """å®å‘½ä»¤"""
    
    def __init__(self, commands):
        self.commands = commands
    
    def execute(self):
        for command in self.commands:
            command.execute()
    
    def undo(self):
        for command in reversed(self.commands):
            command.undo()
    
    def get_description(self):
        descriptions = [cmd.get_description() for cmd in self.commands]
        return f"å®å‘½ä»¤: {', '.join(descriptions)}"

class RemoteControl:
    """é¥æ§å™¨ - è°ƒç”¨è€…"""
    
    def __init__(self):
        self._command = None
        self._command_history = []
    
    def set_command(self, command: Command):
        """è®¾ç½®å‘½ä»¤"""
        self._command = command
        print(f"       è®¾ç½®å‘½ä»¤: {command.get_description()}")
    
    def press_button(self):
        """æŒ‰ä¸‹æŒ‰é’®"""
        if self._command:
            self._command.execute()
            self._command_history.append(self._command)
        else:
            print("       æœªè®¾ç½®å‘½ä»¤")
    
    def press_undo(self):
        """æŒ‰ä¸‹æ’¤é”€æŒ‰é’®"""
        if self._command_history:
            last_command = self._command_history.pop()
            last_command.undo()
        else:
            print("       æ²¡æœ‰å¯æ’¤é”€çš„å‘½ä»¤")
    
    def get_command_history(self):
        return [cmd.get_description() for cmd in self._command_history]

# æµ‹è¯•å‘½ä»¤æ¨¡å¼
print(f"\n   å‘½ä»¤æ¨¡å¼æµ‹è¯•:")

# åˆ›å»ºæ¥æ”¶è€…
living_room_light = Light("å®¢å…")
bedroom_light = Light("å§å®¤")

# åˆ›å»ºå‘½ä»¤
living_room_on = LightOnCommand(living_room_light)
living_room_off = LightOffCommand(living_room_light)
bedroom_on = LightOnCommand(bedroom_light)
bedroom_off = LightOffCommand(bedroom_light)

# åˆ›å»ºå®å‘½ä»¤
all_lights_on = MacroCommand([living_room_on, bedroom_on])
all_lights_off = MacroCommand([living_room_off, bedroom_off])

# åˆ›å»ºè°ƒç”¨è€…
remote = RemoteControl()

# æ‰§è¡Œå•ä¸ªå‘½ä»¤
print(f"\n     å•ä¸ªå‘½ä»¤æµ‹è¯•:")
remote.set_command(living_room_on)
remote.press_button()

remote.set_command(bedroom_on)
remote.press_button()

print(f"\n     å½“å‰çŠ¶æ€:")
print(f"       {living_room_light.get_status()}")
print(f"       {bedroom_light.get_status()}")

# æ‰§è¡Œå®å‘½ä»¤
print(f"\n     å®å‘½ä»¤æµ‹è¯•:")
remote.set_command(all_lights_off)
remote.press_button()

print(f"\n     æ’¤é”€æ“ä½œ:")
remote.press_undo()

print(f"\n     å‘½ä»¤å†å²: {remote.get_command_history()}")

# 4. å·¥å‚æ¨¡å¼ (Factory Pattern)
print(f"\nğŸ“ 4. å·¥å‚æ¨¡å¼ (Factory Pattern):")

class AnimalFactory(ABC):
    """åŠ¨ç‰©å·¥å‚æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def create_animal(self, name):
        """åˆ›å»ºåŠ¨ç‰©"""
        pass

class Dog:
    """ç‹—ç±»"""
    
    def __init__(self, name):
        self.name = name
    
    def make_sound(self):
        return f"{self.name}æ±ªæ±ªå«"
    
    def get_type(self):
        return "ç‹—"

class Cat:
    """çŒ«ç±»"""
    
    def __init__(self, name):
        self.name = name
    
    def make_sound(self):
        return f"{self.name}å–µå–µå«"
    
    def get_type(self):
        return "çŒ«"

class Bird:
    """é¸Ÿç±»"""
    
    def __init__(self, name):
        self.name = name
    
    def make_sound(self):
        return f"{self.name}å•¾å•¾å«"
    
    def get_type(self):
        return "é¸Ÿ"

class SimpleAnimalFactory(AnimalFactory):
    """ç®€å•åŠ¨ç‰©å·¥å‚"""
    
    def create_animal(self, animal_type, name):
        if animal_type.lower() == "dog":
            return Dog(name)
        elif animal_type.lower() == "cat":
            return Cat(name)
        elif animal_type.lower() == "bird":
            return Bird(name)
        else:
            raise ValueError(f"æœªçŸ¥çš„åŠ¨ç‰©ç±»å‹: {animal_type}")

class DogFactory(AnimalFactory):
    """ç‹—å·¥å‚"""
    
    def create_animal(self, name):
        return Dog(name)

class CatFactory(AnimalFactory):
    """çŒ«å·¥å‚"""
    
    def create_animal(self, name):
        return Cat(name)

class AnimalFactoryProducer:
    """å·¥å‚ç”Ÿäº§è€…"""
    
    @staticmethod
    def get_factory(factory_type):
        if factory_type.lower() == "simple":
            return SimpleAnimalFactory()
        elif factory_type.lower() == "dog":
            return DogFactory()
        elif factory_type.lower() == "cat":
            return CatFactory()
        else:
            raise ValueError(f"æœªçŸ¥çš„å·¥å‚ç±»å‹: {factory_type}")

# æµ‹è¯•å·¥å‚æ¨¡å¼
print(f"\n   å·¥å‚æ¨¡å¼æµ‹è¯•:")

# ä½¿ç”¨ç®€å•å·¥å‚
print(f"\n     ç®€å•å·¥å‚æµ‹è¯•:")
simple_factory = AnimalFactoryProducer.get_factory("simple")

animals = []
animals.append(simple_factory.create_animal("dog", "æ—ºè´¢"))
animals.append(simple_factory.create_animal("cat", "å’ªå’ª"))
animals.append(simple_factory.create_animal("bird", "å°é¸Ÿ"))

for animal in animals:
    print(f"       {animal.make_sound()} ({animal.get_type()})")

# ä½¿ç”¨ä¸“é—¨å·¥å‚
print(f"\n     ä¸“é—¨å·¥å‚æµ‹è¯•:")
dog_factory = AnimalFactoryProducer.get_factory("dog")
cat_factory = AnimalFactoryProducer.get_factory("cat")

dogs = []
dogs.append(dog_factory.create_animal("å¤§é»„"))
dogs.append(dog_factory.create_animal("å°ç™½"))

cats = []
cats.append(cat_factory.create_animal("èŠ±èŠ±"))
cats.append(cat_factory.create_animal("é»‘é»‘"))

print(f"\n     ç‹—ç‹—ä»¬:")
for dog in dogs:
    print(f"       {dog.make_sound()}")

print(f"\n     çŒ«å’ªä»¬:")
for cat in cats:
    print(f"       {cat.make_sound()}")

print(f"\nâœ… æ¥å£è®¾è®¡æ¨¡å¼å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å¸¸è§æ¥å£è®¾è®¡æ¨¡å¼çš„å®ç°")
print(f"   âœ“ ç†è§£æ¥å£éš”ç¦»å’Œä¾èµ–å€’ç½®åŸåˆ™")
print(f"   âœ“ èƒ½è®¾è®¡å¯æ‰©å±•çš„ç»„ä»¶æ¶æ„")
````

````python
print("ğŸ—ï¸ å®é™…åº”ç”¨æ¡ˆä¾‹:")
print("=" * 50)

# æ¡ˆä¾‹1: æ•°æ®å¤„ç†ç®¡é“
print(f"ğŸ“ 1. æ•°æ®å¤„ç†ç®¡é“:")

class DataProcessor(ABC):
    """æ•°æ®å¤„ç†å™¨æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def process(self, data):
        """å¤„ç†æ•°æ®"""
        pass
    
    @abstractmethod
    def validate_input(self, data):
        """éªŒè¯è¾“å…¥æ•°æ®"""
        pass
    
    @abstractmethod
    def get_processor_info(self):
        """è·å–å¤„ç†å™¨ä¿¡æ¯"""
        pass
    
    def execute(self, data):
        """æ‰§è¡Œå¤„ç†æµç¨‹"""
        if not self.validate_input(data):
            raise ValueError("è¾“å…¥æ•°æ®éªŒè¯å¤±è´¥")
        return self.process(data)

class DataValidator(ABC):
    """æ•°æ®éªŒè¯å™¨æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def validate(self, data):
        """éªŒè¯æ•°æ®"""
        pass
    
    @abstractmethod
    def get_error_message(self):
        """è·å–é”™è¯¯ä¿¡æ¯"""
        pass

class DataTransformer(ABC):
    """æ•°æ®è½¬æ¢å™¨æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def transform(self, data):
        """è½¬æ¢æ•°æ®"""
        pass
    
    @abstractmethod
    def get_transform_info(self):
        """è·å–è½¬æ¢ä¿¡æ¯"""
        pass

class CSVProcessor(DataProcessor):
    """CSVæ•°æ®å¤„ç†å™¨"""
    
    def __init__(self, delimiter=","):
        self.delimiter = delimiter
        self.processed_count = 0
    
    def validate_input(self, data):
        return isinstance(data, str) and len(data.strip()) > 0
    
    def process(self, data):
        lines = data.strip().split('\n')
        headers = lines[0].split(self.delimiter)
        rows = []
        
        for line in lines[1:]:
            if line.strip():
                values = line.split(self.delimiter)
                row = dict(zip(headers, values))
                rows.append(row)
        
        self.processed_count += len(rows)
        return rows
    
    def get_processor_info(self):
        return f"CSVå¤„ç†å™¨ (åˆ†éš”ç¬¦: {self.delimiter}, å·²å¤„ç†: {self.processed_count}è¡Œ)"

class JSONProcessor(DataProcessor):
    """JSONæ•°æ®å¤„ç†å™¨"""
    
    def __init__(self):
        self.processed_count = 0
    
    def validate_input(self, data):
        if not isinstance(data, str):
            return False
        try:
            import json
            json.loads(data)
            return True
        except json.JSONDecodeError:
            return False
    
    def process(self, data):
        import json
        result = json.loads(data)
        self.processed_count += 1
        return result
    
    def get_processor_info(self):
        return f"JSONå¤„ç†å™¨ (å·²å¤„ç†: {self.processed_count}ä¸ªå¯¹è±¡)"

class DataPipeline:
    """æ•°æ®å¤„ç†ç®¡é“"""
    
    def __init__(self):
        self.processors = []
        self.validators = []
        self.transformers = []
        self.execution_log = []
    
    def add_processor(self, processor: DataProcessor):
        """æ·»åŠ å¤„ç†å™¨"""
        self.processors.append(processor)
        return self
    
    def add_validator(self, validator: DataValidator):
        """æ·»åŠ éªŒè¯å™¨"""
        self.validators.append(validator)
        return self
    
    def add_transformer(self, transformer: DataTransformer):
        """æ·»åŠ è½¬æ¢å™¨"""
        self.transformers.append(transformer)
        return self
    
    def execute(self, data):
        """æ‰§è¡Œç®¡é“"""
        current_data = data
        
        # éªŒè¯é˜¶æ®µ
        for validator in self.validators:
            if not validator.validate(current_data):
                error_msg = validator.get_error_message()
                self.execution_log.append(f"éªŒè¯å¤±è´¥: {error_msg}")
                raise ValueError(error_msg)
        
        # å¤„ç†é˜¶æ®µ
        for processor in self.processors:
            try:
                current_data = processor.execute(current_data)
                self.execution_log.append(f"å¤„ç†æˆåŠŸ: {processor.get_processor_info()}")
            except Exception as e:
                self.execution_log.append(f"å¤„ç†å¤±è´¥: {processor.get_processor_info()} - {e}")
                raise
        
        # è½¬æ¢é˜¶æ®µ
        for transformer in self.transformers:
            try:
                current_data = transformer.transform(current_data)
                self.execution_log.append(f"è½¬æ¢æˆåŠŸ: {transformer.get_transform_info()}")
            except Exception as e:
                self.execution_log.append(f"è½¬æ¢å¤±è´¥: {transformer.get_transform_info()} - {e}")
                raise
        
        return current_data
    
    def get_execution_log(self):
        return self.execution_log.copy()
    
    def clear_log(self):
        self.execution_log.clear()
        return self

# æµ‹è¯•æ•°æ®å¤„ç†ç®¡é“
print(f"   æ•°æ®å¤„ç†ç®¡é“æµ‹è¯•:")

# åˆ›å»ºç®¡é“
pipeline = DataPipeline()

# æ·»åŠ å¤„ç†å™¨
csv_processor = CSVProcessor(",")
json_processor = JSONProcessor()

pipeline.add_processor(csv_processor)

# æµ‹è¯•CSVå¤„ç†
csv_data = """name,age,city
å¼ ä¸‰,25,åŒ—äº¬
æå››,30,ä¸Šæµ·
ç‹äº”,28,å¹¿å·"""

print(f"\n     CSVæ•°æ®å¤„ç†:")
try:
    result = pipeline.execute(csv_data)
    print(f"       å¤„ç†ç»“æœ: {result}")
    print(f"       å¤„ç†å™¨ä¿¡æ¯: {csv_processor.get_processor_info()}")
except Exception as e:
    print(f"       å¤„ç†å¤±è´¥: {e}")

print(f"\n     æ‰§è¡Œæ—¥å¿—:")
for log in pipeline.get_execution_log():
    print(f"       {log}")

# æµ‹è¯•JSONå¤„ç†
pipeline.clear_log()
pipeline.processors.clear()
pipeline.add_processor(json_processor)

json_data = '{"users": [{"name": "å¼ ä¸‰", "age": 25}, {"name": "æå››", "age": 30}]}'

print(f"\n     JSONæ•°æ®å¤„ç†:")
try:
    result = pipeline.execute(json_data)
    print(f"       å¤„ç†ç»“æœ: {result}")
    print(f"       å¤„ç†å™¨ä¿¡æ¯: {json_processor.get_processor_info()}")
except Exception as e:
    print(f"       å¤„ç†å¤±è´¥: {e}")

# æ¡ˆä¾‹2: æ’ä»¶ç³»ç»Ÿæ¶æ„
print(f"\nğŸ“ 2. æ’ä»¶ç³»ç»Ÿæ¶æ„:")

class Plugin(ABC):
    """æ’ä»¶æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def initialize(self):
        """åˆå§‹åŒ–æ’ä»¶"""
        pass
    
    @abstractmethod
    def execute(self, *args, **kwargs):
        """æ‰§è¡Œæ’ä»¶åŠŸèƒ½"""
        pass
    
    @abstractmethod
    def cleanup(self):
        """æ¸…ç†æ’ä»¶èµ„æº"""
        pass
    
    @abstractmethod
    def get_plugin_info(self):
        """è·å–æ’ä»¶ä¿¡æ¯"""
        pass
    
    @property
    @abstractmethod
    def name(self):
        """æ’ä»¶åç§°"""
        pass
    
    @property
    @abstractmethod
    def version(self):
        """æ’ä»¶ç‰ˆæœ¬"""
        pass

class PluginManager:
    """æ’ä»¶ç®¡ç†å™¨"""
    
    def __init__(self):
        self._plugins = {}
        self._loaded_plugins = {}
        self._execution_history = []
    
    def register_plugin(self, plugin: Plugin):
        """æ³¨å†Œæ’ä»¶"""
        if plugin.name in self._plugins:
            print(f"       æ’ä»¶ {plugin.name} å·²å­˜åœ¨ï¼Œå°†è¢«è¦†ç›–")
        
        self._plugins[plugin.name] = plugin
        print(f"       æ’ä»¶ {plugin.name} v{plugin.version} æ³¨å†ŒæˆåŠŸ")
        return True
    
    def unregister_plugin(self, plugin_name):
        """æ³¨é”€æ’ä»¶"""
        if plugin_name in self._plugins:
            if plugin_name in self._loaded_plugins:
                self.unload_plugin(plugin_name)
            del self._plugins[plugin_name]
            print(f"       æ’ä»¶ {plugin_name} æ³¨é”€æˆåŠŸ")
            return True
        return False
    
    def load_plugin(self, plugin_name):
        """åŠ è½½æ’ä»¶"""
        if plugin_name not in self._plugins:
            raise ValueError(f"æ’ä»¶ {plugin_name} æœªæ³¨å†Œ")
        
        if plugin_name in self._loaded_plugins:
            print(f"       æ’ä»¶ {plugin_name} å·²åŠ è½½")
            return self._loaded_plugins[plugin_name]
        
        plugin = self._plugins[plugin_name]
        try:
            plugin.initialize()
            self._loaded_plugins[plugin_name] = plugin
            print(f"       æ’ä»¶ {plugin_name} åŠ è½½æˆåŠŸ")
            return plugin
        except Exception as e:
            print(f"       æ’ä»¶ {plugin_name} åŠ è½½å¤±è´¥: {e}")
            raise
    
    def unload_plugin(self, plugin_name):
        """å¸è½½æ’ä»¶"""
        if plugin_name not in self._loaded_plugins:
            print(f"       æ’ä»¶ {plugin_name} æœªåŠ è½½")
            return False
        
        plugin = self._loaded_plugins[plugin_name]
        try:
            plugin.cleanup()
            del self._loaded_plugins[plugin_name]
            print(f"       æ’ä»¶ {plugin_name} å¸è½½æˆåŠŸ")
            return True
        except Exception as e:
            print(f"       æ’ä»¶ {plugin_name} å¸è½½å¤±è´¥: {e}")
            raise
    
    def execute_plugin(self, plugin_name, *args, **kwargs):
        """æ‰§è¡Œæ’ä»¶"""
        plugin = self.load_plugin(plugin_name)
        
        try:
            result = plugin.execute(*args, **kwargs)
            self._execution_history.append({
                'plugin': plugin_name,
                'args': args,
                'kwargs': kwargs,
                'result': result,
                'success': True
            })
            return result
        except Exception as e:
            self._execution_history.append({
                'plugin': plugin_name,
                'args': args,
                'kwargs': kwargs,
                'error': str(e),
                'success': False
            })
            raise
    
    def get_registered_plugins(self):
        """è·å–å·²æ³¨å†Œæ’ä»¶åˆ—è¡¨"""
        return list(self._plugins.keys())
    
    def get_loaded_plugins(self):
        """è·å–å·²åŠ è½½æ’ä»¶åˆ—è¡¨"""
        return list(self._loaded_plugins.keys())
    
    def get_plugin_info(self, plugin_name):
        """è·å–æ’ä»¶ä¿¡æ¯"""
        if plugin_name in self._plugins:
            return self._plugins[plugin_name].get_plugin_info()
        return None
    
    def get_execution_history(self):
        """è·å–æ‰§è¡Œå†å²"""
        return self._execution_history.copy()

# å…·ä½“æ’ä»¶å®ç°
class LoggerPlugin(Plugin):
    """æ—¥å¿—æ’ä»¶"""
    
    def __init__(self):
        self._logs = []
    
    @property
    def name(self):
        return "logger"
    
    @property
    def version(self):
        return "1.0.0"
    
    def initialize(self):
        print(f"       æ—¥å¿—æ’ä»¶åˆå§‹åŒ–")
    
    def execute(self, message, level="INFO"):
        log_entry = f"[{level}] {message}"
        self._logs.append(log_entry)
        print(f"       æ—¥å¿—è®°å½•: {log_entry}")
        return log_entry
    
    def cleanup(self):
        self._logs.clear()
        print(f"       æ—¥å¿—æ’ä»¶æ¸…ç†å®Œæˆ")
    
    def get_plugin_info(self):
        return {
            "name": self.name,
            "version": self.version,
            "description": "æ—¥å¿—è®°å½•æ’ä»¶",
            "log_count": len(self._logs)
        }
    
    def get_logs(self):
        return self._logs.copy()

class CalculatorPlugin(Plugin):
    """è®¡ç®—å™¨æ’ä»¶"""
    
    def __init__(self):
        self.history = []
    
    @property
    def name(self):
        return "calculator"
    
    @property
    def version(self):
        return "1.0.0"
    
    def initialize(self):
        print(f"       è®¡ç®—å™¨æ’ä»¶åˆå§‹åŒ–")
    
    def execute(self, operation, a, b):
        if operation == "add":
            result = a + b
        elif operation == "subtract":
            result = a - b
        elif operation == "multiply":
            result = a * b
        elif operation == "divide":
            if b == 0:
                raise ValueError("é™¤é›¶é”™è¯¯")
            result = a / b
        else:
            raise ValueError(f"ä¸æ”¯æŒçš„æ“ä½œ: {operation}")
        
        self.history.append(f"{a} {operation} {b} = {result}")
        print(f"       è®¡ç®—ç»“æœ: {a} {operation} {b} = {result}")
        return result
    
    def cleanup(self):
        self.history.clear()
        print(f"       è®¡ç®—å™¨æ’ä»¶æ¸…ç†å®Œæˆ")
    
    def get_plugin_info(self):
        return {
            "name": self.name,
            "version": self.version,
            "description": "æ•°å­¦è®¡ç®—æ’ä»¶",
            "history_count": len(self.history)
        }
    
    def get_history(self):
        return self.history.copy()

class TextProcessorPlugin(Plugin):
    """æ–‡æœ¬å¤„ç†æ’ä»¶"""
    
    def __init__(self):
        self.processed_texts = []
    
    @property
    def name(self):
        return "text_processor"
    
    @property
    def version(self):
        return "1.0.0"
    
    def initialize(self):
        print(f"       æ–‡æœ¬å¤„ç†æ’ä»¶åˆå§‹åŒ–")
    
    def execute(self, text, operation="upper"):
        if operation == "upper":
            result = text.upper()
        elif operation == "lower":
            result = text.lower()
        elif operation == "title":
            result = text.title()
        elif operation == "reverse":
            result = text[::-1]
        elif operation == "word_count":
            result = len(text.split())
        else:
            raise ValueError(f"ä¸æ”¯æŒçš„æ“ä½œ: {operation}")
        
        self.processed_texts.append({
            "original": text,
            "operation": operation,
            "result": result
        })
        
        print(f"       æ–‡æœ¬å¤„ç†: {text} -> {operation} -> {result}")
        return result
    
    def cleanup(self):
        self.processed_texts.clear()
        print(f"       æ–‡æœ¬å¤„ç†æ’ä»¶æ¸…ç†å®Œæˆ")
    
    def get_plugin_info(self):
        return {
            "name": self.name,
            "version": self.version,
            "description": "æ–‡æœ¬å¤„ç†æ’ä»¶",
            "processed_count": len(self.processed_texts)
        }
    
    def get_processed_texts(self):
        return self.processed_texts.copy()

# æµ‹è¯•æ’ä»¶ç³»ç»Ÿ
print(f"\n   æ’ä»¶ç³»ç»Ÿæµ‹è¯•:")

# åˆ›å»ºæ’ä»¶ç®¡ç†å™¨
plugin_manager = PluginManager()

# åˆ›å»ºæ’ä»¶
logger_plugin = LoggerPlugin()
calculator_plugin = CalculatorPlugin()
text_plugin = TextProcessorPlugin()

# æ³¨å†Œæ’ä»¶
plugin_manager.register_plugin(logger_plugin)
plugin_manager.register_plugin(calculator_plugin)
plugin_manager.register_plugin(text_plugin)

print(f"\n     å·²æ³¨å†Œæ’ä»¶: {plugin_manager.get_registered_plugins()}")

# æ‰§è¡Œæ’ä»¶
print(f"\n     æ‰§è¡Œæ’ä»¶æµ‹è¯•:")

# æ—¥å¿—æ’ä»¶
plugin_manager.execute_plugin("logger", "ç³»ç»Ÿå¯åŠ¨", "INFO")
plugin_manager.execute_plugin("logger", "ç”¨æˆ·ç™»å½•", "INFO")

# è®¡ç®—å™¨æ’ä»¶
plugin_manager.execute_plugin("calculator", "add", 10, 20)
plugin_manager.execute_plugin("calculator", "multiply", 5, 6)

# æ–‡æœ¬å¤„ç†æ’ä»¶
plugin_manager.execute_plugin("text_processor", "Hello World", "upper")
plugin_manager.execute_plugin("text_processor", "Python Programming", "title")

print(f"\n     å·²åŠ è½½æ’ä»¶: {plugin_manager.get_loaded_plugins()}")

# è·å–æ’ä»¶ä¿¡æ¯
print(f"\n     æ’ä»¶ä¿¡æ¯:")
for plugin_name in plugin_manager.get_loaded_plugins():
    info = plugin_manager.get_plugin_info(plugin_name)
    print(f"       {plugin_name}: {info}")

# å¸è½½æ’ä»¶
print(f"\n     å¸è½½æ’ä»¶:")
plugin_manager.unload_plugin("calculator")
print(f"       å·²åŠ è½½æ’ä»¶: {plugin_manager.get_loaded_plugins()}")

# æ¡ˆä¾‹3: å­˜å‚¨æŠ½è±¡å±‚
print(f"\nğŸ“ 3. å­˜å‚¨æŠ½è±¡å±‚:")

class StorageInterface(ABC):
    """å­˜å‚¨æ¥å£æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def save(self, key, data):
        """ä¿å­˜æ•°æ®"""
        pass
    
    @abstractmethod
    def load(self, key):
        """åŠ è½½æ•°æ®"""
        pass
    
    @abstractmethod
    def delete(self, key):
        """åˆ é™¤æ•°æ®"""
        pass
    
    @abstractmethod
    def exists(self, key):
        """æ£€æŸ¥æ•°æ®æ˜¯å¦å­˜åœ¨"""
        pass
    
    @abstractmethod
    def list_keys(self):
        """åˆ—å‡ºæ‰€æœ‰é”®"""
        pass
    
    @abstractmethod
    def get_storage_info(self):
        """è·å–å­˜å‚¨ä¿¡æ¯"""
        pass

class MemoryStorage(StorageInterface):
    """å†…å­˜å­˜å‚¨å®ç°"""
    
    def __init__(self):
        self._data = {}
        self._created_at = None
        import time
        self._created_at = time.time()
    
    def save(self, key, data):
        self._data[key] = data
        return True
    
    def load(self, key):
        if key in self._data:
            return self._data[key]
        raise KeyError(f"é”® {key} ä¸å­˜åœ¨")
    
    def delete(self, key):
        if key in self._data:
            del self._data[key]
            return True
        return False
    
    def exists(self, key):
        return key in self._data
    
    def list_keys(self):
        return list(self._data.keys())
    
    def get_storage_info(self):
        import time
        return {
            "type": "MemoryStorage",
            "item_count": len(self._data),
            "created_at": self._created_at,
            "uptime": time.time() - self._created_at
        }

class FileStorage(StorageInterface):
    """æ–‡ä»¶å­˜å‚¨å®ç°"""
    
    def __init__(self, base_path="./storage"):
        self.base_path = base_path
        self._ensure_directory()
    
    def _ensure_directory(self):
        import os
        if not os.path.exists(self.base_path):
            os.makedirs(self.base_path)
    
    def _get_file_path(self, key):
        import os
        return os.path.join(self.base_path, f"{key}.dat")
    
    def save(self, key, data):
        import pickle
        file_path = self._get_file_path(key)
        
        try:
            with open(file_path, 'wb') as f:
                pickle.dump(data, f)
            return True
        except Exception as e:
            print(f"       ä¿å­˜å¤±è´¥: {e}")
            return False
    
    def load(self, key):
        import pickle
        file_path = self._get_file_path(key)
        
        try:
            with open(file_path, 'rb') as f:
                return pickle.load(f)
        except FileNotFoundError:
            raise KeyError(f"é”® {key} ä¸å­˜åœ¨")
        except Exception as e:
            raise RuntimeError(f"åŠ è½½å¤±è´¥: {e}")
    
    def delete(self, key):
        import os
        file_path = self._get_file_path(key)
        
        try:
            os.remove(file_path)
            return True
        except FileNotFoundError:
            return False
        except Exception as e:
            print(f"       åˆ é™¤å¤±è´¥: {e}")
            return False
    
    def exists(self, key):
        import os
        file_path = self._get_file_path(key)
        return os.path.exists(file_path)
    
    def list_keys(self):
        import os
        keys = []
        if os.path.exists(self.base_path):
            for filename in os.listdir(self.base_path):
                if filename.endswith('.dat'):
                    keys.append(filename[:-4])  # ç§»é™¤.datåç¼€
        return keys
    
    def get_storage_info(self):
        import os
        return {
            "type": "FileStorage",
            "base_path": self.base_path,
            "file_count": len(self.list_keys())
        }

class StorageManager:
    """å­˜å‚¨ç®¡ç†å™¨"""
    
    def __init__(self, storage: StorageInterface):
        self.storage = storage
        self.operation_log = []
    
    def save_data(self, key, data):
        """ä¿å­˜æ•°æ®"""
        try:
            result = self.storage.save(key, data)
            self.operation_log.append({
                "operation": "save",
                "key": key,
                "success": True
            })
            return result
        except Exception as e:
            self.operation_log.append({
                "operation": "save",
                "key": key,
                "error": str(e),
                "success": False
            })
            raise
    
    def load_data(self, key):
        """åŠ è½½æ•°æ®"""
        try:
            result = self.storage.load(key)
            self.operation_log.append({
                "operation": "load",
                "key": key,
                "success": True
            })
            return result
        except Exception as e:
            self.operation_log.append({
                "operation": "load",
                "key": key,
                "error": str(e),
                "success": False
            })
            raise
    
    def delete_data(self, key):
        """åˆ é™¤æ•°æ®"""
        try:
            result = self.storage.delete(key)
            self.operation_log.append({
                "operation": "delete",
                "key": key,
                "success": True
            })
            return result
        except Exception as e:
            self.operation_log.append({
                "operation": "delete",
                "key": key,
                "error": str(e),
                "success": False
            })
            raise
    
    def data_exists(self, key):
        """æ£€æŸ¥æ•°æ®æ˜¯å¦å­˜åœ¨"""
        return self.storage.exists(key)
    
    def list_all_keys(self):
        """åˆ—å‡ºæ‰€æœ‰é”®"""
        return self.storage.list_keys()
    
    def get_storage_info(self):
        """è·å–å­˜å‚¨ä¿¡æ¯"""
        return self.storage.get_storage_info()
    
    def get_operation_log(self):
        """è·å–æ“ä½œæ—¥å¿—"""
        return self.operation_log.copy()

# æµ‹è¯•å­˜å‚¨æŠ½è±¡å±‚
print(f"\n   å­˜å‚¨æŠ½è±¡å±‚æµ‹è¯•:")

# å†…å­˜å­˜å‚¨æµ‹è¯•
print(f"\n     å†…å­˜å­˜å‚¨æµ‹è¯•:")
memory_storage = MemoryStorage()
memory_manager = StorageManager(memory_storage)

# ä¿å­˜æ•°æ®
memory_manager.save_data("user1", {"name": "å¼ ä¸‰", "age": 25})
memory_manager.save_data("user2", {"name": "æå››", "age": 30})
memory_manager.save_data("config", {"theme": "dark", "lang": "zh"})

# åŠ è½½æ•°æ®
user1 = memory_manager.load_data("user1")
print(f"       user1: {user1}")

# åˆ—å‡ºæ‰€æœ‰é”®
print(f"       æ‰€æœ‰é”®: {memory_manager.list_all_keys()}")

# å­˜å‚¨ä¿¡æ¯
print(f"       å­˜å‚¨ä¿¡æ¯: {memory_manager.get_storage_info()}")

# æ–‡ä»¶å­˜å‚¨æµ‹è¯•
print(f"\n     æ–‡ä»¶å­˜å‚¨æµ‹è¯•:")
file_storage = FileStorage("./test_storage")
file_manager = StorageManager(file_storage)

# ä¿å­˜æ•°æ®
file_manager.save_data("product1", {"name": "æ‰‹æœº", "price": 2999})
file_manager.save_data("product2", {"name": "ç”µè„‘", "price": 5999})

# åŠ è½½æ•°æ®
product1 = file_manager.load_data("product1")
print(f"       product1: {product1}")

# åˆ—å‡ºæ‰€æœ‰é”®
print(f"       æ‰€æœ‰é”®: {file_manager.list_all_keys()}")

# å­˜å‚¨ä¿¡æ¯
print(f"       å­˜å‚¨ä¿¡æ¯: {file_manager.get_storage_info()}")

print(f"\nâœ… å®é™…åº”ç”¨æ¡ˆä¾‹å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡æŠ½è±¡åŸºç±»åœ¨å®é™…ç³»ç»Ÿä¸­çš„åº”ç”¨")
print(f"   âœ“ ç†è§£æ¥å£è®¾è®¡å¯¹ç³»ç»Ÿæ¶æ„çš„å½±å“")
print(f"   âœ“ èƒ½åº”ç”¨ABCå’Œæ¥å£è®¾è®¡è§£å†³å¤æ‚ä¸šåŠ¡é—®é¢˜")
````
