````python
print("ğŸ”„ å¤šæ€åŸºç¡€:")
print("=" * 50)

# 1. å¤šæ€çš„åŸºæœ¬æ¦‚å¿µ
print(f"ğŸ“ 1. å¤šæ€çš„åŸºæœ¬æ¦‚å¿µ:")

class Animal:
    """åŠ¨ç‰©åŸºç±»"""
    
    def make_sound(self):
        """å‘å‡ºå£°éŸ³ - å¤šæ€æ¥å£"""
        pass
    
    def move(self):
        """ç§»åŠ¨ - å¤šæ€æ¥å£"""
        pass
    
    def eat(self):
        """è¿›é£Ÿ - å¤šæ€æ¥å£"""
        return "æ­£åœ¨è¿›é£Ÿ"

class Dog(Animal):
    """ç‹—ç±»"""
    
    def make_sound(self):
        return "æ±ªæ±ªæ±ª"
    
    def move(self):
        return "å››è¶³å¥”è·‘"
    
    def fetch(self):
        return "å»æ¡çƒ"

class Cat(Animal):
    """çŒ«ç±»"""
    
    def make_sound(self):
        return "å–µå–µå–µ"
    
    def move(self):
        return "ä¼˜é›…è¡Œèµ°"
    
    def climb(self):
        return "çˆ¬æ ‘"

class Bird(Animal):
    """é¸Ÿç±»"""
    
    def make_sound(self):
        return "å•¾å•¾å•¾"
    
    def move(self):
        return "é£ç¿”"
    
    def fly(self):
        return "åœ¨ç©ºä¸­ç¿±ç¿”"

class Fish(Animal):
    """é±¼ç±»"""
    
    def make_sound(self):
        return "åæ³¡æ³¡"
    
    def move(self):
        return "æ¸¸æ³³"
    
    def dive(self):
        return "æ·±æ½œ"

# å¤šæ€å‡½æ•° - æ¥å—ä¸åŒç±»å‹çš„å¯¹è±¡
def animal_sound(animal):
    """è®©åŠ¨ç‰©å‘å‡ºå£°éŸ³ - å¤šæ€åº”ç”¨"""
    return f"{animal.__class__.__name__}: {animal.make_sound()}"

def animal_move(animal):
    """è®©åŠ¨ç‰©ç§»åŠ¨ - å¤šæ€åº”ç”¨"""
    return f"{animal.__class__.__name__}: {animal.move()}"

def animal_action(animal):
    """åŠ¨ç‰©ç»¼åˆè¡Œä¸º - å¤šæ€åº”ç”¨"""
    actions = []
    actions.append(animal.make_sound())
    actions.append(animal.move())
    actions.append(animal.eat())
    return actions

# æµ‹è¯•å¤šæ€
print(f"   å¤šæ€æµ‹è¯•:")
animals = [Dog(), Cat(), Bird(), Fish()]

for animal in animals:
    print(f"\n     {animal.__class__.__name__}:")
    print(f"       å£°éŸ³: {animal.make_sound()}")
    print(f"       ç§»åŠ¨: {animal.move()}")
    print(f"       è¿›é£Ÿ: {animal.eat()}")
    print(f"       å¤šæ€å‡½æ•°è°ƒç”¨: {animal_sound(animal)}")

# 2. é¸­å­ç±»å‹ (Duck Typing)
print(f"\nğŸ“ 2. é¸­å­ç±»å‹ (Duck Typing):")

class Duck:
    """é¸­å­ç±» - ä¸ç»§æ‰¿Animalä½†æœ‰ç›¸åŒæ¥å£"""
    
    def make_sound(self):
        return "å˜å˜å˜"
    
    def move(self):
        return "æ‘‡æ‘†è¡Œèµ°"
    
    def eat(self):
        return "åƒè™«å­"

class Robot:
    """æœºå™¨äººç±» - æ¨¡æ‹ŸåŠ¨ç‰©è¡Œä¸º"""
    
    def make_sound(self):
        return "å“”å“”å“”"
    
    def move(self):
        return "è½®å­æ»šåŠ¨"
    
    def eat(self):
        return "å……ç”µ"

# é¸­å­ç±»å‹å‡½æ•° - ä¸æ£€æŸ¥ç±»å‹ï¼Œåªæ£€æŸ¥æ¥å£
def make_it_sound(obj):
    """è®©å¯¹è±¡å‘å‡ºå£°éŸ³ - é¸­å­ç±»å‹"""
    if hasattr(obj, 'make_sound'):
        return f"{obj.__class__.__name__}: {obj.make_sound()}"
    else:
        return f"{obj.__class__.__name__}: ä¸ä¼šå‘å‡ºå£°éŸ³"

def make_it_move(obj):
    """è®©å¯¹è±¡ç§»åŠ¨ - é¸­å­ç±»å‹"""
    if hasattr(obj, 'move'):
        return f"{obj.__class__.__name__}: {obj.move()}"
    else:
        return f"{obj.__class__.__name__}: ä¸ä¼šç§»åŠ¨"

# æµ‹è¯•é¸­å­ç±»å‹
print(f"   é¸­å­ç±»å‹æµ‹è¯•:")
duck = Duck()
robot = Robot()

duck_typing_objects = [Dog(), Cat(), duck, robot]

for obj in duck_typing_objects:
    print(f"\n     {obj.__class__.__name__}:")
    print(f"       å£°éŸ³: {make_it_sound(obj)}")
    print(f"       ç§»åŠ¨: {make_it_move(obj)}")
    print(f"       æ˜¯Animalå®ä¾‹: {isinstance(obj, Animal)}")

# 3. æŠ½è±¡å¤šæ€æ¥å£
print(f"\nğŸ“ 3. æŠ½è±¡å¤šæ€æ¥å£:")

class Shape:
    """å›¾å½¢æŠ½è±¡åŸºç±»"""
    
    def area(self):
        """è®¡ç®—é¢ç§¯ - æŠ½è±¡æ–¹æ³•"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°areaæ–¹æ³•")
    
    def perimeter(self):
        """è®¡ç®—å‘¨é•¿ - æŠ½è±¡æ–¹æ³•"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°perimeteræ–¹æ³•")
    
    def draw(self):
        """ç»˜åˆ¶å›¾å½¢ - æŠ½è±¡æ–¹æ³•"""
        raise NotImplementedError("å­ç±»å¿…é¡»å®ç°drawæ–¹æ³•")

class Circle(Shape):
    """åœ†å½¢ç±»"""
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
    def perimeter(self):
        return 2 * 3.14159 * self.radius
    
    def draw(self):
        return f"ç»˜åˆ¶åŠå¾„ä¸º{self.radius}çš„åœ†å½¢"

class Rectangle(Shape):
    """çŸ©å½¢ç±»"""
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def draw(self):
        return f"ç»˜åˆ¶å®½{self.width}é«˜{self.height}çš„çŸ©å½¢"

class Triangle(Shape):
    """ä¸‰è§’å½¢ç±»"""
    
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def perimeter(self):
        # ç®€åŒ–ï¼šå‡è®¾ç­‰è…°ä¸‰è§’å½¢
        side = (self.base ** 2 + self.height ** 2) ** 0.5
        return self.base + 2 * side
    
    def draw(self):
        return f"ç»˜åˆ¶åº•{self.base}é«˜{self.height}çš„ä¸‰è§’å½¢"

# å¤šæ€å‡½æ•° - å¤„ç†ä¸åŒå½¢çŠ¶
def calculate_shape_info(shape):
    """è®¡ç®—å½¢çŠ¶ä¿¡æ¯ - å¤šæ€åº”ç”¨"""
    return {
        "type": shape.__class__.__name__,
        "area": shape.area(),
        "perimeter": shape.perimeter(),
        "drawing": shape.draw()
    }

def draw_all_shapes(shapes):
    """ç»˜åˆ¶æ‰€æœ‰å½¢çŠ¶ - å¤šæ€åº”ç”¨"""
    drawings = []
    for shape in shapes:
        drawings.append(shape.draw())
    return drawings

def compare_shapes(shape1, shape2):
    """æ¯”è¾ƒä¸¤ä¸ªå½¢çŠ¶ - å¤šæ€åº”ç”¨"""
    area1 = shape1.area()
    area2 = shape2.area()
    
    if area1 > area2:
        return f"{shape1.__class__.__name__}é¢ç§¯æ›´å¤§"
    elif area2 > area1:
        return f"{shape2.__class__.__name__}é¢ç§¯æ›´å¤§"
    else:
        return "ä¸¤ä¸ªå½¢çŠ¶é¢ç§¯ç›¸ç­‰"

# æµ‹è¯•æŠ½è±¡å¤šæ€æ¥å£
print(f"   æŠ½è±¡å¤šæ€æ¥å£æµ‹è¯•:")
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4)
]

for shape in shapes:
    info = calculate_shape_info(shape)
    print(f"\n     {info['type']}:")
    print(f"       é¢ç§¯: {info['area']:.2f}")
    print(f"       å‘¨é•¿: {info['perimeter']:.2f}")
    print(f"       ç»˜åˆ¶: {info['drawing']}")

print(f"\n   ç»˜åˆ¶æ‰€æœ‰å½¢çŠ¶:")
drawings = draw_all_shapes(shapes)
for drawing in drawings:
    print(f"     {drawing}")

print(f"\n   æ¯”è¾ƒå½¢çŠ¶:")
print(f"     {compare_shapes(shapes[0], shapes[1])}")
print(f"     {compare_shapes(shapes[1], shapes[2])}")

# 4. å¤šæ€ä¸ç»§æ‰¿çš„ç»“åˆ
print(f"\nğŸ“ 4. å¤šæ€ä¸ç»§æ‰¿çš„ç»“åˆ:")

class Vehicle:
    """äº¤é€šå·¥å…·åŸºç±»"""
    
    def start(self):
        """å¯åŠ¨ - å¤šæ€æ¥å£"""
        pass
    
    def stop(self):
        """åœæ­¢ - å¤šæ€æ¥å£"""
        pass
    
    def accelerate(self, speed):
        """åŠ é€Ÿ - å¤šæ€æ¥å£"""
        pass

class Car(Vehicle):
    """æ±½è½¦ç±»"""
    
    def start(self):
        return "æ±½è½¦å¼•æ“å¯åŠ¨"
    
    def stop(self):
        return "æ±½è½¦åˆ¹è½¦åœæ­¢"
    
    def accelerate(self, speed):
        return f"æ±½è½¦åŠ é€Ÿåˆ°{speed}km/h"

class Motorcycle(Vehicle):
    """æ‘©æ‰˜è½¦ç±»"""
    
    def start(self):
        return "æ‘©æ‰˜è½¦å¼•æ“å¯åŠ¨"
    
    def stop(self):
        return "æ‘©æ‰˜è½¦åˆ¹è½¦åœæ­¢"
    
    def accelerate(self, speed):
        return f"æ‘©æ‰˜è½¦åŠ é€Ÿåˆ°{speed}km/h"

class Bicycle(Vehicle):
    """è‡ªè¡Œè½¦ç±»"""
    
    def start(self):
        return "å¼€å§‹è¸©è¸æ¿"
    
    def stop(self):
        return "åœæ­¢è¸©è¸æ¿"
    
    def accelerate(self, speed):
        return f"è‡ªè¡Œè½¦åŠ é€Ÿåˆ°{speed}km/h"

# äº¤é€šç®¡ç†ç³»ç»Ÿ - å¤šæ€åº”ç”¨
class TrafficManager:
    """äº¤é€šç®¡ç†å™¨"""
    
    def __init__(self):
        self.vehicles = []
    
    def add_vehicle(self, vehicle):
        """æ·»åŠ äº¤é€šå·¥å…·"""
        self.vehicles.append(vehicle)
    
    def start_all(self):
        """å¯åŠ¨æ‰€æœ‰è½¦è¾† - å¤šæ€åº”ç”¨"""
        results = []
        for vehicle in self.vehicles:
            results.append(vehicle.start())
        return results
    
    def stop_all(self):
        """åœæ­¢æ‰€æœ‰è½¦è¾† - å¤šæ€åº”ç”¨"""
        results = []
        for vehicle in self.vehicles:
            results.append(vehicle.stop())
        return results
    
    def accelerate_all(self, speed):
        """æ‰€æœ‰è½¦è¾†åŠ é€Ÿ - å¤šæ€åº”ç”¨"""
        results = []
        for vehicle in self.vehicles:
            results.append(vehicle.accelerate(speed))
        return results

# æµ‹è¯•å¤šæ€ä¸ç»§æ‰¿ç»“åˆ
print(f"   å¤šæ€ä¸ç»§æ‰¿ç»“åˆæµ‹è¯•:")
manager = TrafficManager()
manager.add_vehicle(Car())
manager.add_vehicle(Motorcycle())
manager.add_vehicle(Bicycle())

print(f"\n     å¯åŠ¨æ‰€æœ‰è½¦è¾†:")
for result in manager.start_all():
    print(f"       {result}")

print(f"\n     æ‰€æœ‰è½¦è¾†åŠ é€Ÿåˆ°60km/h:")
for result in manager.accelerate_all(60):
    print(f"       {result}")

print(f"\n     åœæ­¢æ‰€æœ‰è½¦è¾†:")
for result in manager.stop_all():
    print(f"       {result}")

print(f"\nâœ… å¤šæ€åŸºç¡€å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å¤šæ€çš„å®ç°æ–¹å¼å’Œä½¿ç”¨åœºæ™¯")
print(f"   âœ“ ç†è§£é¸­å­ç±»å‹çš„æ¦‚å¿µå’Œåº”ç”¨")
print(f"   âœ“ èƒ½è®¾è®¡å¤šæ€æ¥å£å’ŒæŠ½è±¡æ–¹æ³•")
````

````python
print("ğŸ”’ å°è£…åŸºç¡€:")
print("=" * 50)

# 1. å°è£…çš„åŸºæœ¬æ¦‚å¿µ
print(f"ğŸ“ 1. å°è£…çš„åŸºæœ¬æ¦‚å¿µ:")

class BankAccount:
    """é“¶è¡Œè´¦æˆ·ç±» - å°è£…ç¤ºä¾‹"""
    
    def __init__(self, account_number, account_holder, initial_balance=0):
        # å…¬æœ‰å±æ€§
        self.account_number = account_number
        self.account_holder = account_holder
        
        # å—ä¿æŠ¤å±æ€§ (å•ä¸‹åˆ’çº¿çº¦å®š)
        self._balance = initial_balance
        self._transaction_count = 0
        
        # ç§æœ‰å±æ€§ (åŒä¸‹åˆ’çº¿åç§°ä¿®é¥°)
        self.__pin = "0000"
        self.__is_active = True
        self.__daily_limit = 10000
        self.__daily_withdrawn = 0
    
    # å…¬æœ‰æ–¹æ³• - å®‰å…¨æ¥å£
    def deposit(self, amount):
        """å­˜æ¬¾"""
        if amount > 0:
            self._balance += amount
            self._transaction_count += 1
            return f"å­˜æ¬¾æˆåŠŸï¼Œä½™é¢: {self._balance}"
        return "å­˜æ¬¾é‡‘é¢å¿…é¡»å¤§äº0"
    
    def withdraw(self, amount, pin):
        """å–æ¬¾"""
        if not self.__verify_pin(pin):
            return "PINç é”™è¯¯"
        
        if not self.__is_active:
            return "è´¦æˆ·å·²å†»ç»“"
        
        if amount <= 0:
            return "å–æ¬¾é‡‘é¢å¿…é¡»å¤§äº0"
        
        if amount > self._balance:
            return "ä½™é¢ä¸è¶³"
        
        if self.__daily_withdrawn + amount > self.__daily_limit:
            return "è¶…å‡ºæ¯æ—¥å–æ¬¾é™é¢"
        
        self._balance -= amount
        self.__daily_withdrawn += amount
        self._transaction_count += 1
        return f"å–æ¬¾æˆåŠŸï¼Œä½™é¢: {self._balance}"
    
    def get_balance(self):
        """æŸ¥è¯¢ä½™é¢"""
        return self._balance
    
    def get_account_info(self):
        """è·å–è´¦æˆ·ä¿¡æ¯"""
        return {
            "account_number": self.account_number,
            "account_holder": self.account_holder,
            "balance": self._balance,
            "transaction_count": self._transaction_count,
            "is_active": self.__is_active
        }
    
    # å—ä¿æŠ¤æ–¹æ³• - å†…éƒ¨ä½¿ç”¨
    def _log_transaction(self, transaction_type, amount):
        """è®°å½•äº¤æ˜“æ—¥å¿—"""
        print(f"   äº¤æ˜“æ—¥å¿—: {transaction_type} {amount}")
    
    # ç§æœ‰æ–¹æ³• - å†…éƒ¨å®ç°
    def __verify_pin(self, pin):
        """éªŒè¯PINç """
        return pin == self.__pin
    
    def __check_daily_limit(self, amount):
        """æ£€æŸ¥æ¯æ—¥é™é¢"""
        return self.__daily_withdrawn + amount <= self.__daily_limit
    
    def __reset_daily_limit(self):
        """é‡ç½®æ¯æ—¥é™é¢"""
        self.__daily_withdrawn = 0

# æµ‹è¯•å°è£…
print(f"   å°è£…æµ‹è¯•:")
account = BankAccount("123456789", "å¼ ä¸‰", 5000)

print(f"\n     åˆ›å»ºè´¦æˆ·:")
print(f"       {account.get_account_info()}")

print(f"\n     å­˜æ¬¾æ“ä½œ:")
print(f"       {account.deposit(2000)}")
print(f"       {account.deposit(-500)}")

print(f"\n     å–æ¬¾æ“ä½œ:")
print(f"       {account.withdraw(1000, '0000')}")
print(f"       {account.withdraw(5000, '1234')}")
print(f"       {account.withdraw(15000, '0000')}")

print(f"\n     è®¿é—®æ§åˆ¶æ¼”ç¤º:")
print(f"       å…¬æœ‰å±æ€§ account_number: {account.account_number}")
print(f"       å—ä¿æŠ¤å±æ€§ _balance: {account._balance}")
try:
    print(f"       ç§æœ‰å±æ€§ __pin: {account.__pin}")
except AttributeError as e:
    print(f"       ç§æœ‰å±æ€§è®¿é—®å¤±è´¥: {e}")

# 2. å±æ€§å°è£… - ä½¿ç”¨property
print(f"\nğŸ“ 2. å±æ€§å°è£… - ä½¿ç”¨property:")

class Temperature:
    """æ¸©åº¦ç±» - å±æ€§å°è£…ç¤ºä¾‹"""
    
    def __init__(self, celsius=0):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """è·å–æ‘„æ°æ¸©åº¦"""
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        """è®¾ç½®æ‘„æ°æ¸©åº¦"""
        if value < -273.15:
            raise ValueError("æ¸©åº¦ä¸èƒ½ä½äºç»å¯¹é›¶åº¦")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """è·å–åæ°æ¸©åº¦ - è®¡ç®—å±æ€§"""
        return self._celsius * 9/5 + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value):
        """è®¾ç½®åæ°æ¸©åº¦"""
        celsius = (value - 32) * 5/9
        self.celsius = celsius
    
    @property
    def kelvin(self):
        """è·å–å¼€å°”æ–‡æ¸©åº¦ - è®¡ç®—å±æ€§"""
        return self._celsius + 273.15
    
    @property
    def description(self):
        """æ¸©åº¦æè¿° - åªè¯»å±æ€§"""
        if self._celsius < 0:
            return "å†°ç‚¹ä»¥ä¸‹"
        elif self._celsius < 20:
            return "å¯’å†·"
        elif self._celsius < 30:
            return "é€‚å®œ"
        else:
            return "ç‚çƒ­"

class Student:
    """å­¦ç”Ÿç±» - å±æ€§å°è£…ç¤ºä¾‹"""
    
    def __init__(self, name, student_id):
        self.name = name
        self._student_id = student_id
        self._scores = []
        self._attendance = 0
    
    @property
    def student_id(self):
        """è·å–å­¦å· - åªè¯»å±æ€§"""
        return self._student_id
    
    @property
    def scores(self):
        """è·å–æˆç»©åˆ—è¡¨ - åªè¯»å±æ€§"""
        return self._scores.copy()
    
    @property
    def average_score(self):
        """è·å–å¹³å‡åˆ† - è®¡ç®—å±æ€§"""
        if not self._scores:
            return 0
        return sum(self._scores) / len(self._scores)
    
    @property
    def grade(self):
        """è·å–ç­‰çº§ - è®¡ç®—å±æ€§"""
        avg = self.average_score
        if avg >= 90:
            return "A"
        elif avg >= 80:
            return "B"
        elif avg >= 70:
            return "C"
        elif avg >= 60:
            return "D"
        else:
            return "F"
    
    @property
    def attendance_rate(self):
        """è·å–å‡ºå‹¤ç‡ - è®¡ç®—å±æ€§"""
        total_classes = 100  # å‡è®¾æ€»è¯¾ç¨‹æ•°
        return self._attendance / total_classes
    
    def add_score(self, score):
        """æ·»åŠ æˆç»©"""
        if 0 <= score <= 100:
            self._scores.append(score)
        else:
            raise ValueError("æˆç»©å¿…é¡»åœ¨0-100ä¹‹é—´")
    
    def attend_class(self):
        """ä¸Šè¯¾ç­¾åˆ°"""
        self._attendance += 1
    
    def get_summary(self):
        """è·å–å­¦ç”Ÿæ‘˜è¦"""
        return {
            "name": self.name,
            "student_id": self.student_id,
            "average_score": self.average_score,
            "grade": self.grade,
            "attendance_rate": f"{self.attendance_rate:.1%}"
        }

# æµ‹è¯•å±æ€§å°è£…
print(f"   å±æ€§å°è£…æµ‹è¯•:")

# æ¸©åº¦ç±»æµ‹è¯•
temp = Temperature(25)
print(f"\n     æ¸©åº¦ç±»æµ‹è¯•:")
print(f"       æ‘„æ°åº¦: {temp.celsius}Â°C")
print(f"       åæ°åº¦: {temp.fahrenheit}Â°F")
print(f"       å¼€å°”æ–‡: {temp.kelvin}K")
print(f"       æè¿°: {temp.description}")

temp.celsius = 30
print(f"\n     è®¾ç½®ä¸º30Â°Cå:")
print(f"       æ‘„æ°åº¦: {temp.celsius}Â°C")
print(f"       åæ°åº¦: {temp.fahrenheit}Â°F")
print(f"       æè¿°: {temp.description}")

temp.fahrenheit = 86
print(f"\n     è®¾ç½®ä¸º86Â°Få:")
print(f"       æ‘„æ°åº¦: {temp.celsius}Â°C")
print(f"       åæ°åº¦: {temp.fahrenheit}Â°F")

try:
    temp.celsius = -300
except ValueError as e:
    print(f"\n     è®¾ç½®æ— æ•ˆæ¸©åº¦: {e}")

# å­¦ç”Ÿç±»æµ‹è¯•
student = Student("æå››", "2021001")
print(f"\n     å­¦ç”Ÿç±»æµ‹è¯•:")
student.add_score(85)
student.add_score(92)
student.add_score(78)
student.attend_class()
student.attend_class()
student.attend_class()

summary = student.get_summary()
for key, value in summary.items():
    print(f"       {key}: {value}")

# 3. æ–¹æ³•å°è£…
print(f"\nğŸ“ 3. æ–¹æ³•å°è£…:")

class DatabaseConnection:
    """æ•°æ®åº“è¿æ¥ç±» - æ–¹æ³•å°è£…ç¤ºä¾‹"""
    
    def __init__(self, host, port, database):
        self._host = host
        self._port = port
        self._database = database
        self._connection = None
        self._is_connected = False
    
    # å…¬æœ‰æ¥å£
    def connect(self, username, password):
        """è¿æ¥æ•°æ®åº“"""
        if self._is_connected:
            return "å·²ç»è¿æ¥åˆ°æ•°æ®åº“"
        
        if self._validate_credentials(username, password):
            self._connection = self._create_connection()
            self._is_connected = True
            self._log_connection(username)
            return "æ•°æ®åº“è¿æ¥æˆåŠŸ"
        else:
            return "ç”¨æˆ·åæˆ–å¯†ç é”™è¯¯"
    
    def disconnect(self):
        """æ–­å¼€è¿æ¥"""
        if self._is_connected:
            self._close_connection()
            self._is_connected = False
            return "æ•°æ®åº“è¿æ¥å·²æ–­å¼€"
        return "æ•°æ®åº“æœªè¿æ¥"
    
    def execute_query(self, query):
        """æ‰§è¡ŒæŸ¥è¯¢"""
        if not self._is_connected:
            return "è¯·å…ˆè¿æ¥æ•°æ®åº“"
        
        if not self._validate_query(query):
            return "æŸ¥è¯¢è¯­å¥æ— æ•ˆ"
        
        return self._execute_query(query)
    
    def get_connection_info(self):
        """è·å–è¿æ¥ä¿¡æ¯"""
        return {
            "host": self._host,
            "port": self._port,
            "database": self._database,
            "is_connected": self._is_connected
        }
    
    # å—ä¿æŠ¤æ–¹æ³• - å†…éƒ¨ä½¿ç”¨
    def _validate_credentials(self, username, password):
        """éªŒè¯å‡­æ®"""
        # ç®€åŒ–éªŒè¯é€»è¾‘
        return username == "admin" and password == "password"
    
    def _validate_query(self, query):
        """éªŒè¯æŸ¥è¯¢è¯­å¥"""
        # ç®€åŒ–éªŒè¯é€»è¾‘
        return isinstance(query, str) and len(query) > 0
    
    # ç§æœ‰æ–¹æ³• - å†…éƒ¨å®ç°
    def __create_connection(self):
        """åˆ›å»ºè¿æ¥"""
        return f"Connection to {self._host}:{self._port}/{self._database}"
    
    def __close_connection(self):
        """å…³é—­è¿æ¥"""
        self._connection = None
    
    def __execute_query(self, query):
        """æ‰§è¡ŒæŸ¥è¯¢"""
        return f"æ‰§è¡ŒæŸ¥è¯¢: {query}"
    
    def __log_connection(self, username):
        """è®°å½•è¿æ¥æ—¥å¿—"""
        print(f"   æ—¥å¿—: ç”¨æˆ· {username} è¿æ¥åˆ° {self._database}")

# æµ‹è¯•æ–¹æ³•å°è£…
print(f"   æ–¹æ³•å°è£…æµ‹è¯•:")
db = DatabaseConnection("localhost", 5432, "mydb")

print(f"\n     æ•°æ®åº“è¿æ¥æµ‹è¯•:")
print(f"       {db.connect('admin', 'password')}")
print(f"       {db.execute_query('SELECT * FROM users')}")
print(f"       {db.disconnect()}")
print(f"       {db.connect('user', 'wrong')}")

# 4. å°è£…çš„æœ€ä½³å®è·µ
print(f"\nğŸ“ 4. å°è£…çš„æœ€ä½³å®è·µ:")

class SecureData:
    """å®‰å…¨æ•°æ®ç±» - å°è£…æœ€ä½³å®è·µ"""
    
    def __init__(self, data):
        self._data = data
        self._checksum = self._calculate_checksum(data)
        self._is_encrypted = False
    
    @property
    def data(self):
        """è·å–æ•°æ® - åªè¯»å±æ€§"""
        if self._is_encrypted:
            return self._decrypt_data()
        return self._data
    
    @property
    def is_valid(self):
        """æ£€æŸ¥æ•°æ®å®Œæ•´æ€§"""
        current_checksum = self._calculate_checksum(self._data)
        return current_checksum == self._checksum
    
    def update_data(self, new_data, password=None):
        """æ›´æ–°æ•°æ®"""
        if password and not self._verify_password(password):
            raise PermissionError("å¯†ç é”™è¯¯")
        
        self._data = new_data
        self._checksum = self._calculate_checksum(new_data)
        return "æ•°æ®æ›´æ–°æˆåŠŸ"
    
    def encrypt(self, key):
        """åŠ å¯†æ•°æ®"""
        if self._is_encrypted:
            return "æ•°æ®å·²åŠ å¯†"
        
        self._data = self._encrypt_data(self._data, key)
        self._is_encrypted = True
        return "æ•°æ®åŠ å¯†æˆåŠŸ"
    
    def decrypt(self, key):
        """è§£å¯†æ•°æ®"""
        if not self._is_encrypted:
            return "æ•°æ®æœªåŠ å¯†"
        
        self._data = self._decrypt_data(self._data, key)
        self._is_encrypted = False
        return "æ•°æ®è§£å¯†æˆåŠŸ"
    
    # ç§æœ‰æ–¹æ³•
    def _calculate_checksum(self, data):
        """è®¡ç®—æ ¡éªŒå’Œ"""
        return hash(str(data))
    
    def _verify_password(self, password):
        """éªŒè¯å¯†ç """
        return password == "secret123"
    
    def _encrypt_data(self, data, key):
        """åŠ å¯†æ•°æ®"""
        # ç®€åŒ–åŠ å¯†é€»è¾‘
        return f"encrypted_{data}_{key}"
    
    def _decrypt_data(self, data, key=None):
        """è§£å¯†æ•°æ®"""
        # ç®€åŒ–è§£å¯†é€»è¾‘
        if data.startswith("encrypted_"):
            parts = data.split("_")
            return parts[1]
        return data

# æµ‹è¯•å°è£…æœ€ä½³å®è·µ
print(f"   å°è£…æœ€ä½³å®è·µæµ‹è¯•:")
secure_data = SecureData("æ•æ„Ÿä¿¡æ¯")

print(f"\n     æ•°æ®å®Œæ•´æ€§:")
print(f"       æ•°æ®: {secure_data.data}")
print(f"       å®Œæ•´æ€§: {secure_data.is_valid}")

print(f"\n     æ•°æ®åŠ å¯†:")
print(f"       {secure_data.encrypt('mykey')}")
print(f"       åŠ å¯†åæ•°æ®: {secure_data.data}")
print(f"       {secure_data.decrypt('mykey')}")
print(f"       è§£å¯†åæ•°æ®: {secure_data.data}")

print(f"\nâœ… å°è£…åŸºç¡€å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å°è£…çš„åŸºæœ¬æ¦‚å¿µå’ŒåŸåˆ™")
print(f"   âœ“ ç†è§£è®¿é—®æ§åˆ¶çš„æ–¹æ³•å’Œçº¦å®š")
print(f"   âœ“ èƒ½è®¾è®¡å°è£…è‰¯å¥½çš„ç±»ç»“æ„")
````

````python
print("ğŸ—ï¸ å®é™…åº”ç”¨æ¡ˆä¾‹:")
print("=" * 50)

# æ¡ˆä¾‹1: æ”¯ä»˜å¤„ç†ç³»ç»Ÿ
print(f"ğŸ“ 1. æ”¯ä»˜å¤„ç†ç³»ç»Ÿ:")

from abc import ABC, abstractmethod

class PaymentProcessor(ABC):
    """æ”¯ä»˜å¤„ç†å™¨æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def process_payment(self, amount, currency):
        """å¤„ç†æ”¯ä»˜ - å¤šæ€æ¥å£"""
        pass
    
    @abstractmethod
    def refund_payment(self, amount, currency):
        """å¤„ç†é€€æ¬¾ - å¤šæ€æ¥å£"""
        pass
    
    @abstractmethod
    def get_transaction_status(self, transaction_id):
        """è·å–äº¤æ˜“çŠ¶æ€ - å¤šæ€æ¥å£"""
        pass

class CreditCardProcessor(PaymentProcessor):
    """ä¿¡ç”¨å¡æ”¯ä»˜å¤„ç†å™¨"""
    
    def __init__(self, merchant_id, api_key):
        self._merchant_id = merchant_id
        self._api_key = api_key
        self._transactions = {}
    
    def process_payment(self, amount, currency):
        """å¤„ç†ä¿¡ç”¨å¡æ”¯ä»˜"""
        if not self._validate_amount(amount):
            return {"status": "error", "message": "é‡‘é¢æ— æ•ˆ"}
        
        transaction_id = self._generate_transaction_id()
        self._transactions[transaction_id] = {
            "amount": amount,
            "currency": currency,
            "status": "completed",
            "type": "payment"
        }
        
        return {
            "status": "success",
            "transaction_id": transaction_id,
            "message": f"ä¿¡ç”¨å¡æ”¯ä»˜ {amount} {currency} æˆåŠŸ"
        }
    
    def refund_payment(self, amount, currency):
        """å¤„ç†ä¿¡ç”¨å¡é€€æ¬¾"""
        transaction_id = self._generate_transaction_id()
        self._transactions[transaction_id] = {
            "amount": amount,
            "currency": currency,
            "status": "refunded",
            "type": "refund"
        }
        
        return {
            "status": "success",
            "transaction_id": transaction_id,
            "message": f"ä¿¡ç”¨å¡é€€æ¬¾ {amount} {currency} æˆåŠŸ"
        }
    
    def get_transaction_status(self, transaction_id):
        """è·å–äº¤æ˜“çŠ¶æ€"""
        transaction = self._transactions.get(transaction_id)
        if transaction:
            return {
                "status": "success",
                "transaction": transaction
            }
        return {"status": "error", "message": "äº¤æ˜“ä¸å­˜åœ¨"}
    
    def _validate_amount(self, amount):
        """éªŒè¯é‡‘é¢"""
        return amount > 0 and amount <= 100000
    
    def _generate_transaction_id(self):
        """ç”Ÿæˆäº¤æ˜“ID"""
        import random
        return f"CC{random.randint(100000, 999999)}"

class PayPalProcessor(PaymentProcessor):
    """PayPalæ”¯ä»˜å¤„ç†å™¨"""
    
    def __init__(self, client_id, client_secret):
        self._client_id = client_id
        self._client_secret = client_secret
        self._transactions = {}
    
    def process_payment(self, amount, currency):
        """å¤„ç†PayPalæ”¯ä»˜"""
        if not self._validate_currency(currency):
            return {"status": "error", "message": "ä¸æ”¯æŒçš„è´§å¸"}
        
        transaction_id = self._generate_transaction_id()
        self._transactions[transaction_id] = {
            "amount": amount,
            "currency": currency,
            "status": "completed",
            "type": "payment"
        }
        
        return {
            "status": "success",
            "transaction_id": transaction_id,
            "message": f"PayPalæ”¯ä»˜ {amount} {currency} æˆåŠŸ"
        }
    
    def refund_payment(self, amount, currency):
        """å¤„ç†PayPalé€€æ¬¾"""
        transaction_id = self._generate_transaction_id()
        self._transactions[transaction_id] = {
            "amount": amount,
            "currency": currency,
            "status": "refunded",
            "type": "refund"
        }
        
        return {
            "status": "success",
            "transaction_id": transaction_id,
            "message": f"PayPalé€€æ¬¾ {amount} {currency} æˆåŠŸ"
        }
    
    def get_transaction_status(self, transaction_id):
        """è·å–äº¤æ˜“çŠ¶æ€"""
        transaction = self._transactions.get(transaction_id)
        if transaction:
            return {
                "status": "success",
                "transaction": transaction
            }
        return {"status": "error", "message": "äº¤æ˜“ä¸å­˜åœ¨"}
    
    def _validate_currency(self, currency):
        """éªŒè¯è´§å¸"""
        supported_currencies = ["USD", "EUR", "GBP", "CNY"]
        return currency in supported_currencies
    
    def _generate_transaction_id(self):
        """ç”Ÿæˆäº¤æ˜“ID"""
        import random
        return f"PP{random.randint(100000, 999999)}"

class PaymentGateway:
    """æ”¯ä»˜ç½‘å…³ - å¤šæ€å°è£…åº”ç”¨"""
    
    def __init__(self):
        self._processors = {}
        self._default_processor = None
    
    def register_processor(self, name, processor):
        """æ³¨å†Œæ”¯ä»˜å¤„ç†å™¨"""
        if isinstance(processor, PaymentProcessor):
            self._processors[name] = processor
            if not self._default_processor:
                self._default_processor = processor
            return f"æ”¯ä»˜å¤„ç†å™¨ {name} æ³¨å†ŒæˆåŠŸ"
        return "æ— æ•ˆçš„æ”¯ä»˜å¤„ç†å™¨"
    
    def set_default_processor(self, name):
        """è®¾ç½®é»˜è®¤å¤„ç†å™¨"""
        if name in self._processors:
            self._default_processor = self._processors[name]
            return f"é»˜è®¤å¤„ç†å™¨è®¾ç½®ä¸º {name}"
        return "å¤„ç†å™¨ä¸å­˜åœ¨"
    
    def process_payment(self, amount, currency, processor_name=None):
        """å¤„ç†æ”¯ä»˜ - å¤šæ€åº”ç”¨"""
        processor = self._get_processor(processor_name)
        if processor:
            return processor.process_payment(amount, currency)
        return {"status": "error", "message": "æ”¯ä»˜å¤„ç†å™¨ä¸å¯ç”¨"}
    
    def refund_payment(self, amount, currency, processor_name=None):
        """å¤„ç†é€€æ¬¾ - å¤šæ€åº”ç”¨"""
        processor = self._get_processor(processor_name)
        if processor:
            return processor.refund_payment(amount, currency)
        return {"status": "error", "message": "æ”¯ä»˜å¤„ç†å™¨ä¸å¯ç”¨"}
    
    def get_transaction_status(self, transaction_id, processor_name=None):
        """è·å–äº¤æ˜“çŠ¶æ€ - å¤šæ€åº”ç”¨"""
        processor = self._get_processor(processor_name)
        if processor:
            return processor.get_transaction_status(transaction_id)
        return {"status": "error", "message": "æ”¯ä»˜å¤„ç†å™¨ä¸å¯ç”¨"}
    
    def get_available_processors(self):
        """è·å–å¯ç”¨å¤„ç†å™¨åˆ—è¡¨"""
        return list(self._processors.keys())
    
    def _get_processor(self, processor_name):
        """è·å–æ”¯ä»˜å¤„ç†å™¨"""
        if processor_name and processor_name in self._processors:
            return self._processors[processor_name]
        return self._default_processor

# æµ‹è¯•æ”¯ä»˜å¤„ç†ç³»ç»Ÿ
print(f"   æ”¯ä»˜å¤„ç†ç³»ç»Ÿæµ‹è¯•:")
gateway = PaymentGateway()

# æ³¨å†Œæ”¯ä»˜å¤„ç†å™¨
credit_processor = CreditCardProcessor("merchant123", "api_key_456")
paypal_processor = PayPalProcessor("client_id_789", "client_secret_012")

print(f"     {gateway.register_processor('credit_card', credit_processor)}")
print(f"     {gateway.register_processor('paypal', paypal_processor)}")
print(f"     å¯ç”¨å¤„ç†å™¨: {gateway.get_available_processors()}")

# å¤„ç†æ”¯ä»˜
print(f"\n     å¤„ç†æ”¯ä»˜:")
result1 = gateway.process_payment(100, "USD")
print(f"       é»˜è®¤å¤„ç†å™¨: {result1}")

result2 = gateway.process_payment(50, "CNY", "paypal")
print(f"       PayPalå¤„ç†å™¨: {result2}")

result3 = gateway.process_payment(200, "EUR", "credit_card")
print(f"       ä¿¡ç”¨å¡å¤„ç†å™¨: {result3}")

# å¤„ç†é€€æ¬¾
print(f"\n     å¤„ç†é€€æ¬¾:")
refund1 = gateway.refund_payment(25, "USD", "credit_card")
print(f"       {refund1}")

refund2 = gateway.refund_payment(30, "CNY", "paypal")
print(f"       {refund2}")

# æ¡ˆä¾‹2: æ–‡ä»¶å¤„ç†ç³»ç»Ÿ
print(f"\nğŸ“ 2. æ–‡ä»¶å¤„ç†ç³»ç»Ÿ:")

class FileHandler(ABC):
    """æ–‡ä»¶å¤„ç†å™¨æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def read_file(self, filepath):
        """è¯»å–æ–‡ä»¶ - å¤šæ€æ¥å£"""
        pass
    
    @abstractmethod
    def write_file(self, filepath, content):
        """å†™å…¥æ–‡ä»¶ - å¤šæ€æ¥å£"""
        pass
    
    @abstractmethod
    def append_file(self, filepath, content):
        """è¿½åŠ æ–‡ä»¶ - å¤šæ€æ¥å£"""
        pass
    
    @abstractmethod
    def delete_file(self, filepath):
        """åˆ é™¤æ–‡ä»¶ - å¤šæ€æ¥å£"""
        pass

class TextFileHandler(FileHandler):
    """æ–‡æœ¬æ–‡ä»¶å¤„ç†å™¨"""
    
    def __init__(self, encoding="utf-8"):
        self._encoding = encoding
        self._files = {}  # æ¨¡æ‹Ÿæ–‡ä»¶å­˜å‚¨
    
    def read_file(self, filepath):
        """è¯»å–æ–‡æœ¬æ–‡ä»¶"""
        if filepath in self._files:
            return {
                "status": "success",
                "content": self._files[filepath],
                "encoding": self._encoding
            }
        return {"status": "error", "message": "æ–‡ä»¶ä¸å­˜åœ¨"}
    
    def write_file(self, filepath, content):
        """å†™å…¥æ–‡æœ¬æ–‡ä»¶"""
        if not isinstance(content, str):
            return {"status": "error", "message": "å†…å®¹å¿…é¡»æ˜¯å­—ç¬¦ä¸²"}
        
        self._files[filepath] = content
        return {"status": "success", "message": f"æ–‡ä»¶ {filepath} å†™å…¥æˆåŠŸ"}
    
    def append_file(self, filepath, content):
        """è¿½åŠ æ–‡æœ¬æ–‡ä»¶"""
        if filepath in self._files:
            if not isinstance(content, str):
                return {"status": "error", "message": "å†…å®¹å¿…é¡»æ˜¯å­—ç¬¦ä¸²"}
            
            self._files[filepath] += content
            return {"status": "success", "message": f"æ–‡ä»¶ {filepath} è¿½åŠ æˆåŠŸ"}
        return {"status": "error", "message": "æ–‡ä»¶ä¸å­˜åœ¨"}
    
    def delete_file(self, filepath):
        """åˆ é™¤æ–‡æœ¬æ–‡ä»¶"""
        if filepath in self._files:
            del self._files[filepath]
            return {"status": "success", "message": f"æ–‡ä»¶ {filepath} åˆ é™¤æˆåŠŸ"}
        return {"status": "error", "message": "æ–‡ä»¶ä¸å­˜åœ¨"}
    
    def list_files(self):
        """åˆ—å‡ºæ‰€æœ‰æ–‡ä»¶"""
        return list(self._files.keys())

class JSONFileHandler(FileHandler):
    """JSONæ–‡ä»¶å¤„ç†å™¨"""
    
    def __init__(self):
        self._files = {}  # æ¨¡æ‹Ÿæ–‡ä»¶å­˜å‚¨
    
    def read_file(self, filepath):
        """è¯»å–JSONæ–‡ä»¶"""
        if filepath in self._files:
            try:
                import json
                content = json.loads(self._files[filepath])
                return {
                    "status": "success",
                    "content": content,
                    "type": "json"
                }
            except json.JSONDecodeError:
                return {"status": "error", "message": "JSONæ ¼å¼é”™è¯¯"}
        return {"status": "error", "message": "æ–‡ä»¶ä¸å­˜åœ¨"}
    
    def write_file(self, filepath, content):
        """å†™å…¥JSONæ–‡ä»¶"""
        try:
            import json
            json_str = json.dumps(content, ensure_ascii=False, indent=2)
            self._files[filepath] = json_str
            return {"status": "success", "message": f"JSONæ–‡ä»¶ {filepath} å†™å…¥æˆåŠŸ"}
        except (TypeError, ValueError) as e:
            return {"status": "error", "message": f"JSONåºåˆ—åŒ–å¤±è´¥: {e}"}
    
    def append_file(self, filepath, content):
        """JSONæ–‡ä»¶ä¸æ”¯æŒè¿½åŠ """
        return {"status": "error", "message": "JSONæ–‡ä»¶ä¸æ”¯æŒè¿½åŠ æ“ä½œ"}
    
    def delete_file(self, filepath):
        """åˆ é™¤JSONæ–‡ä»¶"""
        if filepath in self._files:
            del self._files[filepath]
            return {"status": "success", "message": f"JSONæ–‡ä»¶ {filepath} åˆ é™¤æˆåŠŸ"}
        return {"status": "error", "message": "æ–‡ä»¶ä¸å­˜åœ¨"}

class FileManager:
    """æ–‡ä»¶ç®¡ç†å™¨ - å¤šæ€å°è£…åº”ç”¨"""
    
    def __init__(self):
        self._handlers = {}
        self._default_handler = None
    
    def register_handler(self, file_extension, handler):
        """æ³¨å†Œæ–‡ä»¶å¤„ç†å™¨"""
        if isinstance(handler, FileHandler):
            self._handlers[file_extension] = handler
            if not self._default_handler:
                self._default_handler = handler
            return f"æ–‡ä»¶å¤„ç†å™¨ {file_extension} æ³¨å†ŒæˆåŠŸ"
        return "æ— æ•ˆçš„æ–‡ä»¶å¤„ç†å™¨"
    
    def get_handler(self, filepath):
        """è·å–æ–‡ä»¶å¤„ç†å™¨"""
        extension = filepath.split('.')[-1] if '.' in filepath else ''
        return self._handlers.get(extension, self._default_handler)
    
    def read_file(self, filepath):
        """è¯»å–æ–‡ä»¶ - å¤šæ€åº”ç”¨"""
        handler = self.get_handler(filepath)
        if handler:
            return handler.read_file(filepath)
        return {"status": "error", "message": "æ²¡æœ‰åˆé€‚çš„æ–‡ä»¶å¤„ç†å™¨"}
    
    def write_file(self, filepath, content):
        """å†™å…¥æ–‡ä»¶ - å¤šæ€åº”ç”¨"""
        handler = self.get_handler(filepath)
        if handler:
            return handler.write_file(filepath, content)
        return {"status": "error", "message": "æ²¡æœ‰åˆé€‚çš„æ–‡ä»¶å¤„ç†å™¨"}
    
    def append_file(self, filepath, content):
        """è¿½åŠ æ–‡ä»¶ - å¤šæ€åº”ç”¨"""
        handler = self.get_handler(filepath)
        if handler:
            return handler.append_file(filepath, content)
        return {"status": "error", "message": "æ²¡æœ‰åˆé€‚çš„æ–‡ä»¶å¤„ç†å™¨"}
    
    def delete_file(self, filepath):
        """åˆ é™¤æ–‡ä»¶ - å¤šæ€åº”ç”¨"""
        handler = self.get_handler(filepath)
        if handler:
            return handler.delete_file(filepath)
        return {"status": "error", "message": "æ²¡æœ‰åˆé€‚çš„æ–‡ä»¶å¤„ç†å™¨"}

# æµ‹è¯•æ–‡ä»¶å¤„ç†ç³»ç»Ÿ
print(f"\n   æ–‡ä»¶å¤„ç†ç³»ç»Ÿæµ‹è¯•:")
file_manager = FileManager()

# æ³¨å†Œæ–‡ä»¶å¤„ç†å™¨
text_handler = TextFileHandler()
json_handler = JSONFileHandler()

print(f"     {file_manager.register_handler('txt', text_handler)}")
print(f"     {file_manager.register_handler('json', json_handler)}")

# å¤„ç†æ–‡æœ¬æ–‡ä»¶
print(f"\n     å¤„ç†æ–‡æœ¬æ–‡ä»¶:")
write_result = file_manager.write_file("document.txt", "è¿™æ˜¯ä¸€ä¸ªæ–‡æœ¬æ–‡ä»¶")
print(f"       {write_result}")

read_result = file_manager.read_file("document.txt")
print(f"       {read_result}")

append_result = file_manager.append_file("document.txt", "\nè¿½åŠ çš„å†…å®¹")
print(f"       {append_result}")

read_result2 = file_manager.read_file("document.txt")
print(f"       {read_result2}")

# å¤„ç†JSONæ–‡ä»¶
print(f"\n     å¤„ç†JSONæ–‡ä»¶:")
json_data = {"name": "å¼ ä¸‰", "age": 25, "city": "åŒ—äº¬"}
write_json = file_manager.write_file("data.json", json_data)
print(f"       {write_json}")

read_json = file_manager.read_file("data.json")
print(f"       {read_json}")

append_json = file_manager.append_file("data.json", "æ–°å†…å®¹")
print(f"       {append_json}")

# æ¡ˆä¾‹3: é€šçŸ¥ç³»ç»Ÿ
print(f"\nğŸ“ 3. é€šçŸ¥ç³»ç»Ÿ:")

class Notifier(ABC):
    """é€šçŸ¥å™¨æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def send_notification(self, recipient, message):
        """å‘é€é€šçŸ¥ - å¤šæ€æ¥å£"""
        pass
    
    @abstractmethod
    def get_delivery_status(self, notification_id):
        """è·å–å‘é€çŠ¶æ€ - å¤šæ€æ¥å£"""
        pass

class EmailNotifier(Notifier):
    """é‚®ä»¶é€šçŸ¥å™¨"""
    
    def __init__(self, smtp_server, smtp_port):
        self._smtp_server = smtp_server
        self._smtp_port = smtp_port
        self._notifications = {}
    
    def send_notification(self, recipient, message):
        """å‘é€é‚®ä»¶é€šçŸ¥"""
        if not self._validate_email(recipient):
            return {"status": "error", "message": "é‚®ç®±åœ°å€æ— æ•ˆ"}
        
        notification_id = self._generate_notification_id()
        self._notifications[notification_id] = {
            "recipient": recipient,
            "message": message,
            "status": "sent",
            "type": "email"
        }
        
        return {
            "status": "success",
            "notification_id": notification_id,
            "message": f"é‚®ä»¶å‘é€è‡³ {recipient}"
        }
    
    def get_delivery_status(self, notification_id):
        """è·å–é‚®ä»¶å‘é€çŠ¶æ€"""
        notification = self._notifications.get(notification_id)
        if notification:
            return {
                "status": "success",
                "notification": notification
            }
        return {"status": "error", "message": "é€šçŸ¥ä¸å­˜åœ¨"}
    
    def _validate_email(self, email):
        """éªŒè¯é‚®ç®±åœ°å€"""
        return "@" in email and "." in email
    
    def _generate_notification_id(self):
        """ç”Ÿæˆé€šçŸ¥ID"""
        import random
        return f"EMAIL{random.randint(100000, 999999)}"

class SMSNotifier(Notifier):
    """çŸ­ä¿¡é€šçŸ¥å™¨"""
    
    def __init__(self, sms_gateway):
        self._sms_gateway = sms_gateway
        self._notifications = {}
    
    def send_notification(self, recipient, message):
        """å‘é€çŸ­ä¿¡é€šçŸ¥"""
        if not self._validate_phone(recipient):
            return {"status": "error", "message": "æ‰‹æœºå·æ— æ•ˆ"}
        
        if len(message) > 160:
            return {"status": "error", "message": "çŸ­ä¿¡å†…å®¹è¿‡é•¿"}
        
        notification_id = self._generate_notification_id()
        self._notifications[notification_id] = {
            "recipient": recipient,
            "message": message,
            "status": "sent",
            "type": "sms"
        }
        
        return {
            "status": "success",
            "notification_id": notification_id,
            "message": f"çŸ­ä¿¡å‘é€è‡³ {recipient}"
        }
    
    def get_delivery_status(self, notification_id):
        """è·å–çŸ­ä¿¡å‘é€çŠ¶æ€"""
        notification = self._notifications.get(notification_id)
        if notification:
            return {
                "status": "success",
                "notification": notification
            }
        return {"status": "error", "message": "é€šçŸ¥ä¸å­˜åœ¨"}
    
    def _validate_phone(self, phone):
        """éªŒè¯æ‰‹æœºå·"""
        return phone.isdigit() and len(phone) == 11
    
    def _generate_notification_id(self):
        """ç”Ÿæˆé€šçŸ¥ID"""
        import random
        return f"SMS{random.randint(100000, 999999)}"

class NotificationService:
    """é€šçŸ¥æœåŠ¡ - å¤šæ€å°è£…åº”ç”¨"""
    
    def __init__(self):
        self._notifiers = {}
        self._default_notifier = None
    
    def register_notifier(self, name, notifier):
        """æ³¨å†Œé€šçŸ¥å™¨"""
        if isinstance(notifier, Notifier):
            self._notifiers[name] = notifier
            if not self._default_notifier:
                self._default_notifier = notifier
            return f"é€šçŸ¥å™¨ {name} æ³¨å†ŒæˆåŠŸ"
        return "æ— æ•ˆçš„é€šçŸ¥å™¨"
    
    def send_notification(self, recipient, message, notifier_type=None):
        """å‘é€é€šçŸ¥ - å¤šæ€åº”ç”¨"""
        notifier = self._get_notifier(notifier_type)
        if notifier:
            return notifier.send_notification(recipient, message)
        return {"status": "error", "message": "é€šçŸ¥å™¨ä¸å¯ç”¨"}
    
    def broadcast_notification(self, recipients, message, notifier_type=None):
        """å¹¿æ’­é€šçŸ¥ - å¤šæ€åº”ç”¨"""
        notifier = self._get_notifier(notifier_type)
        if notifier:
            results = []
            for recipient in recipients:
                result = notifier.send_notification(recipient, message)
                results.append(result)
            return results
        return {"status": "error", "message": "é€šçŸ¥å™¨ä¸å¯ç”¨"}
    
    def get_delivery_status(self, notification_id, notifier_type=None):
        """è·å–å‘é€çŠ¶æ€ - å¤šæ€åº”ç”¨"""
        notifier = self._get_notifier(notifier_type)
        if notifier:
            return notifier.get_delivery_status(notification_id)
        return {"status": "error", "message": "é€šçŸ¥å™¨ä¸å¯ç”¨"}
    
    def _get_notifier(self, notifier_type):
        """è·å–é€šçŸ¥å™¨"""
        if notifier_type and notifier_type in self._notifiers:
            return self._notifiers[notifier_type]
        return self._default_notifier

# æµ‹è¯•é€šçŸ¥ç³»ç»Ÿ
print(f"\n   é€šçŸ¥ç³»ç»Ÿæµ‹è¯•:")
notification_service = NotificationService()

# æ³¨å†Œé€šçŸ¥å™¨
email_notifier = EmailNotifier("smtp.gmail.com", 587)
sms_notifier = SMSNotifier("sms.gateway.com")

print(f"     {notification_service.register_notifier('email', email_notifier)}")
print(f"     {notification_service.register_notifier('sms', sms_notifier)}")

# å‘é€é€šçŸ¥
print(f"\n     å‘é€é€šçŸ¥:")
email_result = notification_service.send_notification(
    "user@example.com", "æ¬¢è¿ä½¿ç”¨æˆ‘ä»¬çš„æœåŠ¡ï¼", "email"
)
print(f"       {email_result}")

sms_result = notification_service.send_notification(
    "13800138000", "æ‚¨çš„éªŒè¯ç æ˜¯123456", "sms"
)
print(f"       {sms_result}")

# å¹¿æ’­é€šçŸ¥
print(f"\n     å¹¿æ’­é€šçŸ¥:")
recipients = ["user1@example.com", "user2@example.com", "user3@example.com"]
broadcast_results = notification_service.broadcast_notification(
    recipients, "ç³»ç»Ÿç»´æŠ¤é€šçŸ¥", "email"
)
for result in broadcast_results:
    print(f"       {result}")

print(f"\nâœ… å®é™…åº”ç”¨æ¡ˆä¾‹å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å¤šæ€å’Œå°è£…åœ¨å®é™…ç³»ç»Ÿä¸­çš„åº”ç”¨")
print(f"   âœ“ ç†è§£æ¥å£è®¾è®¡å’Œå®ç°çš„é‡è¦æ€§")
print(f"   âœ“ èƒ½åº”ç”¨å¤šæ€å’Œå°è£…è§£å†³å¤æ‚ä¸šåŠ¡é—®é¢˜")
````
