```python
print("🌟 开源贡献与社区参与:")
print("=" * 50)

# 导入必要的库
import os
import sys
import re
import json
import datetime
import time
import hashlib
import typing
import dataclasses
import uuid
import inspect
import pathlib
import subprocess
import requests
from typing import List, Dict, Any, Optional, Union, Tuple, Set, Callable
from dataclasses import dataclass, field, asdict
from enum import Enum
from abc import ABC, abstractmethod
from pathlib import Path

print(f"✅ Python版本: {__import__('sys').version}")
print(f"✅ 当前时间: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# 1. 开源贡献管理系统
print(f"📝 1. 开源贡献管理系统:")

# 1.1 贡献类型和角色
print(f"\n   🎯 1.1 贡献类型和角色:")

class ContributionType(Enum):
    """贡献类型"""
    CODE_CONTRIBUTION = "code_contribution"  # 代码贡献
    DOCUMENTATION = "documentation"  # 文档贡献
    BUG_REPORT = "bug_report"  # 错误报告
    FEATURE_REQUEST = "feature_request"  # 功能请求
    TRANSLATION = "translation"  # 翻译贡献
    DESIGN = "design"  # 设计贡献
    TESTING = "testing"  # 测试贡献
    COMMUNITY_SUPPORT = "community_support"  # 社区支持
    SECURITY = "security"  # 安全贡献
    PERFORMANCE = "performance"  # 性能贡献

class ContributorRole(Enum):
    """贡献者角色"""
    OWNER = "owner"  # 项目所有者
    MAINTAINER = "maintainer"  # 维护者
    COMMITTER = "committer"  # 提交者
    CONTRIBUTOR = "contributor"  # 贡献者
    REVIEWER = "reviewer"  # 审查者
    TRIAGER = "triager"  # 问题分类者
    DOCUMENTER = "documenter"  # 文档维护者
    COMMUNITY_MANAGER = "community_manager"  # 社区管理者

@dataclass
class Contributor:
    """贡献者"""
    id: str
    username: str
    email: str
    name: str
    role: ContributorRole
    contributions: List[str]  # 贡献ID列表
    skills: List[str]
    location: str
    timezone: str
    github_url: str
    joined_at: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
    last_active: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
    
    def get_contribution_count(self) -> int:
        """获取贡献数量"""
        return len(self.contributions)
    
    def is_active(self, days: int = 30) -> bool:
        """检查是否活跃"""
        return (datetime.datetime.utcnow() - self.last_active).days <= days

@dataclass
class Contribution:
    """贡献"""
    id: str
    type: ContributionType
    title: str
    description: str
    contributor_id: str
    status: str  # open, in_review, merged, closed, rejected
    priority: int  # 1-10
    created_at: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
    updated_at: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
    reviewed_by: Optional[str] = None
    merged_at: Optional[datetime.datetime] = None
    tags: List[str] = field(default_factory=list)
    metadata: Dict[str, Any] = field(default_factory=dict)
    
    def update_status(self, new_status: str, reviewer_id: Optional[str] = None):
        """更新状态"""
        self.status = new_status
        self.updated_at = datetime.datetime.utcnow()
        
        if reviewer_id:
            self.reviewed_by = reviewer_id
        
        if new_status == "merged":
            self.merged_at = datetime.datetime.utcnow()

@dataclass
class OpenSourceProject:
    """开源项目"""
    id: str
    name: str
    description: str
    repository_url: str
    license: str
    language: str
    owner_id: str
    maintainers: List[str] = field(default_factory=list)
    contributors: List[str] = field(default_factory=list)
    contributions: List[str] = field(default_factory=list)
    created_at: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
    tags: List[str] = field(default_factory=list)
    
    def add_contributor(self, contributor_id: str):
        """添加贡献者"""
        if contributor_id not in self.contributors:
            self.contributors.append(contributor_id)
    
    def add_contribution(self, contribution_id: str):
        """添加贡献"""
        if contribution_id not in self.contributions:
            self.contributions.append(contribution_id)
    
    def get_contributor_stats(self) -> Dict[str, int]:
        """获取贡献者统计"""
        return {
            "total_contributors": len(self.contributors),
            "total_contributions": len(self.contributions),
            "maintainers": len(self.maintainers)
        }

class OpenSourceManager:
    """开源项目管理器"""
    
    def __init__(self):
        self.projects: Dict[str, OpenSourceProject] = {}
        self.contributors: Dict[str, Contributor] = {}
        self.contributions: Dict[str, Contribution] = {}
        self.community_metrics: Dict[str, Any] = {}
        self.contribution_guidelines = {}
        self.code_of_conduct = ""
    
    def create_project(self, project: OpenSourceProject):
        """创建开源项目"""
        self.projects[project.id] = project
        print(f"开源项目创建: {project.name}")
    
    def add_contributor(self, contributor: Contributor, project_id: str):
        """添加贡献者到项目"""
        self.contributors[contributor.id] = contributor
        
        if project_id in self.projects:
            self.projects[project_id].add_contributor(contributor.id)
        
        print(f"贡献者添加: {contributor.username} -> {project_id}")
    
    def submit_contribution(self, contribution: Contribution, project_id: str) -> str:
        """提交贡献"""
        self.contributions[contribution.id] = contribution
        
        if project_id in self.projects:
            self.projects[project_id].add_contribution(contribution.id)
        
        # 更新贡献者信息
        if contribution.contributor_id in self.contributors:
            contributor = self.contributors[contribution.contributor_id]
            contributor.contributions.append(contribution.id)
            contributor.last_active = datetime.datetime.utcnow()
        
        print(f"贡献提交: {contribution.title} ({contribution.type.value})")
        return contribution.id
    
    def review_contribution(self, contribution_id: str, reviewer_id: str, 
                            decision: str, feedback: str = "") -> bool:
        """审查贡献"""
        if contribution_id not in self.contributions:
            return False
        
        contribution = self.contributions[contribution_id]
        
        if decision == "approve":
            contribution.update_status("merged", reviewer_id)
            print(f"贡献批准: {contribution.title}")
        elif decision == "request_changes":
            contribution.update_status("in_review", reviewer_id)
            contribution.metadata["feedback"] = feedback
            print(f"贡献需要修改: {contribution.title}")
        elif decision == "reject":
            contribution.update_status("rejected", reviewer_id)
            contribution.metadata["rejection_reason"] = feedback
            print(f"贡献拒绝: {contribution.title}")
        else:
            return False
        
        return True
    
    def get_project_contributions(self, project_id: str, 
                                  contribution_type: Optional[ContributionType] = None) -> List[Contribution]:
        """获取项目贡献"""
        if project_id not in self.projects:
            return []
        
        project = self.projects[project_id]
        contributions = []
        
        for contrib_id in project.contributions:
            if contrib_id in self.contributions:
                contribution = self.contributions[contrib_id]
                
                if contribution_type is None or contribution.type == contribution_type:
                    contributions.append(contribution)
        
        return contributions
    
    def get_contributor_ranking(self, project_id: str, 
                                time_period: int = 30) -> List[Dict[str, Any]]:
        """获取贡献者排名"""
        contributions = self.get_project_contributions(project_id)
        
        # 统计贡献者贡献
        contributor_stats = {}
        cutoff_date = datetime.datetime.utcnow() - datetime.timedelta(days=time_period)
        
        for contribution in contributions:
            if contribution.created_at >= cutoff_date:
                contributor_id = contribution.contributor_id
                
                if contributor_id not in contributor_stats:
                    contributor_stats[contributor_id] = {
                        "contributor_id": contributor_id,
                        "contributions": 0,
                        "by_type": {}
                        "merged": 0
                    }
                
                stats = contributor_stats[contributor_id]
                stats["contributions"] += 1
                
                # 按类型统计
                contrib_type = contribution.type.value
                stats["by_type"][contrib_type] = stats["by_type"].get(contrib_type, 0) + 1
                
                # 统计合并的贡献
                if contribution.status == "merged":
                    stats["merged"] += 1
        
        # 添加贡献者信息
        for contributor_id, stats in contributor_stats.items():
            if contributor_id in self.contributors:
                contributor = self.contributors[contributor_id]
                stats["username"] = contributor.username
                stats["name"] = contributor.name
                stats["role"] = contributor.role.value
        
        # 排序
        ranking = sorted(contributor_stats.values(), 
                        key=lambda x: (x["merged"], x["contributions"]), 
                        reverse=True)
        
        return ranking
    
    def generate_contribution_report(self, project_id: str) -> Dict[str, Any]:
        """生成贡献报告"""
        if project_id not in self.projects:
            return {}
        
        project = self.projects[project_id]
        contributions = self.get_project_contributions(project_id)
        
        # 按类型统计
        type_stats = {}
        status_stats = {}
        monthly_stats = {}
        
        for contribution in contributions:
            # 类型统计
            contrib_type = contribution.type.value
            type_stats[contrib_type] = type_stats.get(contrib_type, 0) + 1
            
            # 状态统计
            status = contribution.status
            status_stats[status] = status_stats.get(status, 0) + 1
            
            # 月度统计
            month_key = contribution.created_at.strftime("%Y-%m")
            if month_key not in monthly_stats:
                monthly_stats[month_key] = 0
            monthly_stats[month_key] += 1
        
        # 计算合并率
        total_contributions = len(contributions)
        merged_contributions = status_stats.get("merged", 0)
        merge_rate = merged_contributions / total_contributions if total_contributions > 0 else 0
        
        return {
            "project_name": project.name,
            "total_contributions": total_contributions,
            "total_contributors": len(project.contributors),
            "merge_rate": merge_rate,
            "type_distribution": type_stats,
            "status_distribution": status_stats,
            "monthly_trend": monthly_stats,
            "top_contributors": self.get_contributor_ranking(project_id)[:10]
        }
    
    def setup_contribution_guidelines(self, project_id: str, guidelines: Dict[str, str]):
        """设置贡献指南"""
        self.contribution_guidelines[project_id] = guidelines
        print(f"贡献指南设置: {project_id}")
    
    def setup_code_of_conduct(self, code_of_conduct: str):
        """设置行为准则"""
        self.code_of_conduct = code_of_conduct
        print("行为准则设置完成")

print(f"   ✅ 开源贡献管理系统完成")
print(f"      - OpenSourceManager: 开源项目管理器")
print(f"      - 支持10种贡献类型")
print(f"      - 提供8种贡献者角色")
print(f"      - 包含完整的贡献审查流程")

print(f"\n✅ 开源贡献管理系统完成")
print(f"🎯 学习目标达成:")
print(f"   ✓ 掌握开源贡献的流程和规范")
print(f"   ✓ 学会贡献者管理和激励")
print(f"   ✓ 理解开源项目治理结构")
print(f"   ✓ 能够建立完整的开源协作体系")
```

```python
print("\n🏘️ 社区参与与运营管理:")
print("=" * 50)

# 2. 社区运营管理系统
print(f"📝 2. 社区运营管理系统:")

# 2.1 社区活动和参与
print(f"\n   🎯 2.1 社区活动和参与:")

class CommunityEventType(Enum):
    """社区活动类型"""
    HACKATHON = "hackathon"  # 黑客松
    WORKSHOP = "workshop"  # 工作坊
    WEBINAR = "webinar"  # 网络研讨会
    MEETUP = "meetup"  # 聚会
    CONFERENCE = "conference"  # 会议
    CODE_CHALLENGE = "code_challenge"  # 代码挑战
    DOCUMENTATION_SPRINT = "documentation_sprint"  # 文档冲刺
    BUG_BASH = "bug_bash"  # 错误修复活动
    RELEASE_PARTY = "release_party"  # 发布派对
    COMMUNITY_SURVEY = "community_survey"  # 社区调查

class EngagementLevel(Enum):
    """参与度级别"""
    NEWCOMER = "newcomer"  # 新人
    OCCASIONAL = "occasional"  # 偶尔参与
    REGULAR = "regular"  # 常规参与
    ACTIVE = "active"  # 活跃
    CORE = "core"  # 核心成员
    CHAMPION = "champion"  # 倡导者

@dataclass
class CommunityMember:
    """社区成员"""
    id: str
    username: str
    email: str
    name: str
    engagement_level: EngagementLevel
    join_date: datetime.datetime
    last_active: datetime.datetime
    contributions_count: int = 0
    events_attended: int = 0
    reputation_score: int = 0
    badges: List[str] = field(default_factory=list)
    interests: List[str] = field(default_factory=list)
    skills: List[str] = field(default_factory=list)
    
    def update_engagement(self, activity_type: str, points: int = 1):
        """更新参与度"""
        self.last_active = datetime.datetime.utcnow()
        self.reputation_score += points
        
        # 更新参与度级别
        if self.reputation_score >= 1000:
            self.engagement_level = EngagementLevel.CHAMPION
        elif self.reputation_score >= 500:
            self.engagement_level = EngagementLevel.CORE
        elif self.reputation_score >= 200:
            self.engagement_level = EngagementLevel.ACTIVE
        elif self.reputation_score >= 50:
            self.engagement_level = EngagementLevel.REGULAR
        elif self.reputation_score >= 10:
            self.engagement_level = EngagementLevel.OCCASIONAL
    
    def add_badge(self, badge: str):
        """添加徽章"""
        if badge not in self.badges:
            self.badges.append(badge)
            self.reputation_score += 50

@dataclass
class CommunityEvent:
    """社区活动"""
    id: str
    title: str
    description: str
    event_type: CommunityEventType
    start_time: datetime.datetime
    end_time: datetime.datetime
    location: str  # 线上/线下地址
    max_participants: int
    current_participants: int = 0
    participants: List[str] = field(default_factory=list)
    organizers: List[str] = field(default_factory=list)
    status: str = "upcoming"  # upcoming, ongoing, completed, cancelled
    tags: List[str] = field(default_factory=list)
    prizes: List[str] = field(default_factory=list)
    
    def register_participant(self, member_id: str) -> bool:
        """注册参与者"""
        if (self.current_participants < self.max_participants and 
            member_id not in self.participants):
            self.participants.append(member_id)
            self.current_participants += 1
            return True
        return False
    
    def is_full(self) -> bool:
        """检查是否满员"""
        return self.current_participants >= self.max_participants

@dataclass
class DiscussionThread:
    """讨论主题"""
    id: str
    title: str
    content: str
    author_id: str
    category: str
    tags: List[str]
    created_at: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
    last_reply_at: datetime.datetime = field(default_factory=datetime.datetime.utcnow)
    replies: List[Dict[str, Any]] = field(default_factory=list)
    upvotes: int = 0
    downvotes: int = 0
    views: int = 0
    is_pinned: bool = False
    is_locked: bool = False
    
    def add_reply(self, author_id: str, content: str) -> str:
        """添加回复"""
        reply_id = str(uuid.uuid4())
        reply = {
            "id": reply_id,
            "author_id": author_id,
            "content": content,
            "created_at": datetime.datetime.utcnow().isoformat(),
            "upvotes": 0,
            "downvotes": 0
        }
        
        self.replies.append(reply)
        self.last_reply_at = datetime.datetime.utcnow()
        
        return reply_id

class CommunityManager:
    """社区管理器"""
    
    def __init__(self):
        self.members: Dict[str, CommunityMember] = {}
        self.events: Dict[str, CommunityEvent] = {}
        self.discussions: Dict[str, DiscussionThread] = {}
        self.announcements: List[Dict[str, Any]] = []
        self.badges: Dict[str, Dict[str, Any]] = {}
        self.engagement_metrics: Dict[str, Any] = {}
        self.moderation_logs: List[Dict[str, Any]] = []
    
    def register_member(self, member: CommunityMember):
        """注册社区成员"""
        self.members[member.id] = member
        print(f"社区成员注册: {member.username}")
    
    def create_event(self, event: CommunityEvent):
        """创建社区活动"""
        self.events[event.id] = event
        print(f"社区活动创建: {event.title}")
    
    def register_for_event(self, event_id: str, member_id: str) -> bool:
        """注册活动"""
        if event_id not in self.events or member_id not in self.members:
            return False
        
        event = self.events[event_id]
        member = self.members[member_id]
        
        if event.register_participant(member_id):
            member.events_attended += 1
            member.update_engagement("event_registration", 10)
            print(f"活动注册成功: {member.username} -> {event.title}")
            return True
        else:
            print(f"活动注册失败: {event.title} 已满员")
            return False
    
    def create_discussion(self, discussion: DiscussionThread):
        """创建讨论主题"""
        self.discussions[discussion.id] = discussion
        
        # 更新作者参与度
        if discussion.author_id in self.members:
            author = self.members[discussion.author_id]
            author.update_engagement("discussion_created", 20)
        
        print(f"讨论主题创建: {discussion.title}")
    
    def reply_to_discussion(self, discussion_id: str, member_id: str, content: str) -> bool:
        """回复讨论"""
        if discussion_id not in self.discussions or member_id not in self.members:
            return False
        
        discussion = self.discussions[discussion_id]
        member = self.members[member_id]
        
        if not discussion.is_locked:
            reply_id = discussion.add_reply(member_id, content)
            member.update_engagement("discussion_reply", 5)
            
            print(f"讨论回复: {member.username} -> {discussion.title}")
            return True
        
        return False
    
    def upvote_discussion(self, discussion_id: str, member_id: str) -> bool:
        """点赞讨论"""
        if discussion_id not in self.discussions or member_id not in self.members:
            return False
        
        discussion = self.discussions[discussion_id]
        member = self.members[member_id]
        
        discussion.upvotes += 1
        member.update_engagement("discussion_upvote", 2)
        
        # 给作者加分
        if discussion.author_id in self.members:
            author = self.members[discussion.author_id]
            author.update_engagement("discussion_received_upvote", 5)
        
        return True
    
    def award_badge(self, member_id: str, badge_type: str, reason: str) -> bool:
        """颁发徽章"""
        if member_id not in self.members:
            return False
        
        member = self.members[member_id]
        badge_name = f"{badge_type}_{datetime.datetime.utcnow().strftime('%Y%m')}"
        
        member.add_badge(badge_name)
        
        # 记录徽章颁发
        award_record = {
            "member_id": member_id,
            "badge_name": badge_name,
            "reason": reason,
            "awarded_at": datetime.datetime.utcnow().isoformat()
        }
        
        if badge_name not in self.badges:
            self.badges[badge_name] = []
        self.badges[badge_name].append(award_record)
        
        print(f"徽章颁发: {badge_name} -> {member.username}")
        return True
    
    def create_announcement(self, title: str, content: str, 
                           priority: str = "normal", target_audience: List[str] = None):
        """创建公告"""
        announcement = {
            "id": str(uuid.uuid4()),
            "title": title,
            "content": content,
            "priority": priority,
            "target_audience": target_audience or [],
            "created_at": datetime.datetime.utcnow().isoformat(),
            "views": 0
        }
        
        self.announcements.append(announcement)
        print(f"公告发布: {title}")
    
    def get_trending_discussions(self, days: int = 7, limit: int = 10) -> List[Dict[str, Any]]:
        """获取热门讨论"""
        cutoff_date = datetime.datetime.utcnow() - datetime.timedelta(days=days)
        
        trending = []
        for discussion in self.discussions.values():
            if discussion.last_reply_at >= cutoff_date:
                # 计算热度分数
                score = (discussion.upvotes * 3 + 
                        len(discussion.replies) * 2 + 
                        discussion.views * 0.1)
                
                trending.append({
                    "id": discussion.id,
                    "title": discussion.title,
                    "category": discussion.category,
                    "score": score,
                    "replies": len(discussion.replies),
                    "upvotes": discussion.upvotes,
                    "views": discussion.views
                })
        
        # 按热度排序
        trending.sort(key=lambda x: x["score"], reverse=True)
        
        return trending[:limit]
    
    def get_member_leaderboard(self, metric: str = "reputation", limit: int = 20) -> List[Dict[str, Any]]:
        """获取成员排行榜"""
        leaderboard = []
        
        for member in self.members.values():
            if metric == "reputation":
                score = member.reputation_score
            elif metric == "contributions":
                score = member.contributions_count
            elif metric == "events":
                score = member.events_attended
            else:
                score = member.reputation_score
            
            leaderboard.append({
                "member_id": member.id,
                "username": member.username,
                "name": member.name,
                "engagement_level": member.engagement_level.value,
                "score": score,
                "badges": len(member.badges)
            })
        
        # 排序
        leaderboard.sort(key=lambda x: x["score"], reverse=True)
        
        return leaderboard[:limit]
    
    def get_community_health_metrics(self) -> Dict[str, Any]:
        """获取社区健康指标"""
        total_members = len(self.members)
        
        # 活跃成员统计
        active_members = len([
            m for m in self.members.values() 
            if m.is_active(30)
        ])
        
        # 参与度分布
        engagement_distribution = {}
        for member in self.members.values():
            level = member.engagement_level.value
            engagement_distribution[level] = engagement_distribution.get(level, 0) + 1
        
        # 活动统计
        upcoming_events = len([e for e in self.events.values() if e.status == "upcoming"])
        ongoing_events = len([e for e in self.events.values() if e.status == "ongoing"])
        
        # 讨论统计
        total_discussions = len(self.discussions)
        total_replies = sum(len(d.replies) for d in self.discussions.values())
        
        # 计算健康分数
        health_score = 0
        if total_members > 0:
            activity_rate = active_members / total_members
            discussion_rate = total_discussions / total_members
            event_participation = sum(e.current_participants for e in self.events.values()) / total_members
            
            health_score = (activity_rate * 0.4 + discussion_rate * 0.3 + event_participation * 0.3) * 100
        
        return {
            "total_members": total_members,
            "active_members": active_members,
            "activity_rate": active_members / total_members if total_members > 0 else 0,
            "engagement_distribution": engagement_distribution,
            "upcoming_events": upcoming_events,
            "ongoing_events": ongoing_events,
            "total_discussions": total_discussions,
            "total_replies": total_replies,
            "health_score": round(health_score, 2),
            "total_badges_awarded": sum(len(m.badges) for m in self.members.values())
        }
    
    def generate_community_report(self) -> Dict[str, Any]:
        """生成社区报告"""
        health_metrics = self.get_community_health_metrics()
        leaderboard = self.get_member_leaderboard()
        trending_discussions = self.get_trending_discussions()
        
        # 月度增长统计
        monthly_growth = self._calculate_monthly_growth()
        
        return {
            "health_metrics": health_metrics,
            "top_contributors": leaderboard[:10],
            "trending_discussions": trending_discussions[:5],
            "monthly_growth": monthly_growth,
            "recent_announcements": self.announcements[-5:],
            "engagement_trends": self._calculate_engagement_trends()
        }
    
    def _calculate_monthly_growth(self) -> Dict[str, int]:
        """计算月度增长"""
        # 简化实现，实际应该基于历史数据
        current_month = datetime.datetime.utcnow().strftime("%Y-%m")
        
        return {
            "new_members_this_month": 5,  # 示例数据
            "new_discussions_this_month": 12,
            "events_this_month": 3,
            "badges_awarded_this_month": 8
        }
    
    def _calculate_engagement_trends(self) -> Dict[str, List[float]]:
        """计算参与度趋势"""
        # 简化实现，实际应该基于历史数据
        return {
            "daily_active_users": [45, 52, 48, 61, 58, 72, 69],
            "daily_discussions": [3, 5, 4, 7, 6, 8, 9],
            "daily_replies": [12, 18, 15, 22, 19, 28, 25]
        }

print(f"   ✅ 社区运营管理系统完成")
print(f"      - CommunityManager: 社区管理器")
print(f"      - 支持10种社区活动类型")
print(f"      - 提供6种参与度级别")
print(f"      - 包含完整的激励和徽章系统")

print(f"\n✅ 社区参与与运营管理完成")
print(f"🎯 学习目标达成:")
print(f"   ✓ 掌握社区运营的策略和方法")
print(f"   ✓ 学会用户参与和互动管理")
print(f"   ✓ 理解社区治理和决策机制")
print(f"   ✓ 能够建立完整的社区管理体系")
```
