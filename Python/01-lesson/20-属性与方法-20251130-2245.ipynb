````python
print("ğŸ¨ å±æ€§è£…é¥°å™¨:")
print("=" * 50)

# 1. åŸºç¡€å±æ€§è£…é¥°å™¨
print(f"ğŸ“ 1. åŸºç¡€å±æ€§è£…é¥°å™¨:")

class Person:
    """äººç±»ï¼šæ¼”ç¤ºåŸºç¡€å±æ€§è£…é¥°å™¨"""
    
    def __init__(self, name, age):
        self._name = name  # å—ä¿æŠ¤å±æ€§
        self._age = age    # å—ä¿æŠ¤å±æ€§
    
    @property
    def name(self):
        """å§“åå±æ€§ï¼šåªè¯»"""
        return self._name
    
    @property
    def age(self):
        """å¹´é¾„å±æ€§ï¼šåªè¯»"""
        return self._age
    
    @age.setter
    def age(self, new_age):
        """å¹´é¾„è®¾ç½®å™¨ï¼šå¸¦éªŒè¯"""
        if not isinstance(new_age, int):
            raise TypeError("å¹´é¾„å¿…é¡»æ˜¯æ•´æ•°")
        if new_age < 0 or new_age > 150:
            raise ValueError("å¹´é¾„å¿…é¡»åœ¨0-150ä¹‹é—´")
        self._age = new_age
        print(f"   å¹´é¾„å·²æ›´æ–°ä¸º: {new_age}")
    
    @age.deleter
    def age(self):
        """å¹´é¾„åˆ é™¤å™¨"""
        print(f"   åˆ é™¤å¹´é¾„å±æ€§")
        del self._age

# æµ‹è¯•å±æ€§è£…é¥°å™¨
person = Person("å¼ ä¸‰", 25)

print(f"   å±æ€§è®¿é—®æµ‹è¯•:")
print(f"     å§“å: {person.name}")
print(f"     å¹´é¾„: {person.age}")

# ä¿®æ”¹å±æ€§
print(f"\n   å±æ€§ä¿®æ”¹æµ‹è¯•:")
person.age = 30
print(f"     æ–°å¹´é¾„: {person.age}")

# æµ‹è¯•éªŒè¯
print(f"\n   å±æ€§éªŒè¯æµ‹è¯•:")
try:
    person.age = -5
except ValueError as e:
    print(f"     æ•è·åˆ°é¢„æœŸé”™è¯¯: {e}")

try:
    person.age = "äºŒåäº”"
except TypeError as e:
    print(f"     æ•è·åˆ°é¢„æœŸé”™è¯¯: {e}")

# 2. è®¡ç®—å±æ€§
print(f"\nğŸ“ 2. è®¡ç®—å±æ€§:")

class Rectangle:
    """çŸ©å½¢ç±»ï¼šæ¼”ç¤ºè®¡ç®—å±æ€§"""
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    @property
    def area(self):
        """é¢ç§¯ï¼šè®¡ç®—å±æ€§"""
        return self.width * self.height
    
    @property
    def perimeter(self):
        """å‘¨é•¿ï¼šè®¡ç®—å±æ€§"""
        return 2 * (self.width + self.height)
    
    @property
    def diagonal(self):
        """å¯¹è§’çº¿ï¼šè®¡ç®—å±æ€§"""
        return (self.width ** 2 + self.height ** 2) ** 0.5
    
    @property
    def is_square(self):
        """æ˜¯å¦ä¸ºæ­£æ–¹å½¢ï¼šè®¡ç®—å±æ€§"""
        return self.width == self.height

rect = Rectangle(4, 3)
print(f"   çŸ©å½¢å±æ€§æµ‹è¯•:")
print(f"     å®½: {rect.width}, é«˜: {rect.height}")
print(f"     é¢ç§¯: {rect.area}")
print(f"     å‘¨é•¿: {rect.perimeter}")
print(f"     å¯¹è§’çº¿: {rect.diagonal:.2f}")
print(f"     æ˜¯å¦ä¸ºæ­£æ–¹å½¢: {rect.is_square}")

# 3. åªè¯»å±æ€§å’Œç§æœ‰å±æ€§
print(f"\nğŸ“ 3. åªè¯»å±æ€§å’Œç§æœ‰å±æ€§:")

class BankAccount:
    """é“¶è¡Œè´¦æˆ·ï¼šæ¼”ç¤ºåªè¯»å±æ€§"""
    
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # ç§æœ‰å±æ€§
        self.__balance = initial_balance       # ç§æœ‰å±æ€§
        self.__transaction_count = 0          # ç§æœ‰å±æ€§
    
    @property
    def account_number(self):
        """è´¦æˆ·å·ï¼šåªè¯»å±æ€§"""
        return self.__account_number
    
    @property
    def balance(self):
        """ä½™é¢ï¼šåªè¯»å±æ€§"""
        return self.__balance
    
    @property
    def transaction_count(self):
        """äº¤æ˜“æ¬¡æ•°ï¼šåªè¯»å±æ€§"""
        return self.__transaction_count
    
    def deposit(self, amount):
        """å­˜æ¬¾ï¼šä¿®æ”¹ç§æœ‰å±æ€§"""
        if amount > 0:
            self.__balance += amount
            self.__transaction_count += 1
            print(f"   å­˜æ¬¾ {amount} æˆåŠŸï¼Œä½™é¢: {self.__balance}")
        else:
            print(f"   å­˜æ¬¾é‡‘é¢å¿…é¡»å¤§äº0")
    
    def withdraw(self, amount):
        """å–æ¬¾ï¼šä¿®æ”¹ç§æœ‰å±æ€§"""
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            self.__transaction_count += 1
            print(f"   å–æ¬¾ {amount} æˆåŠŸï¼Œä½™é¢: {self.__balance}")
        else:
            print(f"   å–æ¬¾å¤±è´¥ï¼šä½™é¢ä¸è¶³æˆ–é‡‘é¢æ— æ•ˆ")

account = BankAccount("ACC001", 1000)
print(f"   é“¶è¡Œè´¦æˆ·æµ‹è¯•:")
print(f"     è´¦æˆ·å·: {account.account_number}")
print(f"     ä½™é¢: {account.balance}")
print(f"     äº¤æ˜“æ¬¡æ•°: {account.transaction_count}")

account.deposit(500)
account.withdraw(200)
print(f"\n   äº¤æ˜“å:")
print(f"     ä½™é¢: {account.balance}")
print(f"     äº¤æ˜“æ¬¡æ•°: {account.transaction_count}")

# 4. å¤æ‚å±æ€§éªŒè¯
print(f"\nğŸ“ 4. å¤æ‚å±æ€§éªŒè¯:")

class Student:
    """å­¦ç”Ÿç±»ï¼šæ¼”ç¤ºå¤æ‚å±æ€§éªŒè¯"""
    
    def __init__(self, name, student_id):
        self.__name = name
        self.__student_id = student_id
        self.__scores = {}
    
    @property
    def name(self):
        """å§“åï¼šåªè¯»"""
        return self.__name
    
    @property
    def student_id(self):
        """å­¦å·ï¼šåªè¯»"""
        return self.__student_id
    
    @property
    def scores(self):
        """æˆç»©ï¼šåªè¯»å‰¯æœ¬"""
        return self.__scores.copy()
    
    @property
    def average_score(self):
        """å¹³å‡åˆ†ï¼šè®¡ç®—å±æ€§"""
        if not self.__scores:
            return 0
        return sum(self.__scores.values()) / len(self.__scores)
    
    @property
    def grade_level(self):
        """ç­‰çº§ï¼šè®¡ç®—å±æ€§"""
        avg = self.average_score
        if avg >= 90:
            return "ä¼˜ç§€"
        elif avg >= 80:
            return "è‰¯å¥½"
        elif avg >= 70:
            return "ä¸­ç­‰"
        elif avg >= 60:
            return "åŠæ ¼"
        else:
            return "ä¸åŠæ ¼"
    
    def add_score(self, subject, score):
        """æ·»åŠ æˆç»©ï¼šå¸¦éªŒè¯"""
        if not isinstance(score, (int, float)):
            raise TypeError("æˆç»©å¿…é¡»æ˜¯æ•°å­—")
        if score < 0 or score > 100:
            raise ValueError("æˆç»©å¿…é¡»åœ¨0-100ä¹‹é—´")
        if not isinstance(subject, str) or not subject.strip():
            raise TypeError("ç§‘ç›®åç§°å¿…é¡»æ˜¯éç©ºå­—ç¬¦ä¸²")
        
        self.__scores[subject] = score
        print(f"   {self.__name} çš„ {subject} æˆç»©: {score}")
    
    def remove_score(self, subject):
        """åˆ é™¤æˆç»©"""
        if subject in self.__scores:
            del self.__scores[subject]
            print(f"   åˆ é™¤ {subject} æˆç»©")
        else:
            print(f"   æœªæ‰¾åˆ° {subject} æˆç»©")

student = Student("æå››", "S001")
print(f"   å­¦ç”Ÿæˆç»©ç®¡ç†æµ‹è¯•:")
print(f"     å­¦ç”Ÿ: {student.name}, å­¦å·: {student.student_id}")

# æ·»åŠ æˆç»©
student.add_score("æ•°å­¦", 85)
student.add_score("è‹±è¯­", 92)
student.add_score("ç‰©ç†", 78)

print(f"\n   æˆç»©ç»Ÿè®¡:")
print(f"     å„ç§‘æˆç»©: {student.scores}")
print(f"     å¹³å‡åˆ†: {student.average_score:.2f}")
print(f"     ç­‰çº§: {student.grade_level}")

# 5. å±æ€§è£…é¥°å™¨é«˜çº§ç”¨æ³•
print(f"\nğŸ“ 5. å±æ€§è£…é¥°å™¨é«˜çº§ç”¨æ³•:")

class Temperature:
    """æ¸©åº¦ç±»ï¼šæ¼”ç¤ºå±æ€§è£…é¥°å™¨é«˜çº§ç”¨æ³•"""
    
    def __init__(self, celsius=0):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """æ‘„æ°åº¦"""
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        """è®¾ç½®æ‘„æ°åº¦"""
        if value < -273.15:
            raise ValueError("æ¸©åº¦ä¸èƒ½ä½äºç»å¯¹é›¶åº¦")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """åæ°åº¦ï¼šè®¡ç®—å±æ€§"""
        return self._celsius * 9/5 + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value):
        """è®¾ç½®åæ°åº¦ï¼šè‡ªåŠ¨è½¬æ¢ä¸ºæ‘„æ°åº¦"""
        celsius = (value - 32) * 5/9
        self.celsius = celsius  # ä½¿ç”¨celsiusçš„setterè¿›è¡ŒéªŒè¯
    
    @property
    def kelvin(self):
        """å¼€å°”æ–‡ï¼šè®¡ç®—å±æ€§"""
        return self._celsius + 273.15
    
    @kelvin.setter
    def kelvin(self, value):
        """è®¾ç½®å¼€å°”æ–‡ï¼šè‡ªåŠ¨è½¬æ¢ä¸ºæ‘„æ°åº¦"""
        celsius = value - 273.15
        self.celsius = celsius  # ä½¿ç”¨celsiusçš„setterè¿›è¡ŒéªŒè¯

temp = Temperature(25)
print(f"   æ¸©åº¦è½¬æ¢æµ‹è¯•:")
print(f"     æ‘„æ°åº¦: {temp.celsius}Â°C")
print(f"     åæ°åº¦: {temp.fahrenheit:.2f}Â°F")
print(f"     å¼€å°”æ–‡: {temp.kelvin:.2f}K")

# é€šè¿‡ä¸åŒå•ä½è®¾ç½®æ¸©åº¦
print(f"\n   æ¸©åº¦è®¾ç½®æµ‹è¯•:")
temp.fahrenheit = 100
print(f"     è®¾ç½®ä¸º100Â°Få:")
print(f"       æ‘„æ°åº¦: {temp.celsius:.2f}Â°C")
print(f"       åæ°åº¦: {temp.fahrenheit:.2f}Â°F")

temp.kelvin = 373.15
print(f"\n     è®¾ç½®ä¸º373.15Kå:")
print(f"       æ‘„æ°åº¦: {temp.celsius:.2f}Â°C")
print(f"       å¼€å°”æ–‡: {temp.kelvin:.2f}K")

print(f"\nâœ… å±æ€§è£…é¥°å™¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡propertyè£…é¥°å™¨ä½¿ç”¨æ–¹æ³•")
print(f"   âœ“ ç†è§£getterã€setterã€deleterä½œç”¨")
print(f"   âœ“ èƒ½å®ç°æ•°æ®éªŒè¯å’Œè®¡ç®—å±æ€§")
````
