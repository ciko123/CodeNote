````python
print("ğŸ—ï¸ åˆ›å»ºå‹è®¾è®¡æ¨¡å¼:")
print("=" * 50)

# 1. å•ä¾‹æ¨¡å¼ (Singleton Pattern)
print(f"ğŸ“ 1. å•ä¾‹æ¨¡å¼ (Singleton Pattern):")

class SingletonMeta(type):
    """å•ä¾‹å…ƒç±»"""
    
    _instances = {}
    
    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            instance = super().__call__(*args, **kwargs)
            cls._instances[cls] = instance
        return cls._instances[cls]

class DatabaseConnection(metaclass=SingletonMeta):
    """æ•°æ®åº“è¿æ¥å•ä¾‹ç±»"""
    
    def __init__(self):
        self.connection_id = id(self)
        self.is_connected = False
        print(f"   åˆ›å»ºæ•°æ®åº“è¿æ¥å®ä¾‹: {self.connection_id}")
    
    def connect(self, connection_string):
        if not self.is_connected:
            print(f"   è¿æ¥åˆ°æ•°æ®åº“: {connection_string}")
            self.is_connected = True
            self.connection_string = connection_string
        else:
            print(f"   å·²è¿æ¥åˆ°: {self.connection_string}")
    
    def disconnect(self):
        if self.is_connected:
            print(f"   æ–­å¼€æ•°æ®åº“è¿æ¥")
            self.is_connected = False
    
    def execute_query(self, query):
        if not self.is_connected:
            raise RuntimeError("æ•°æ®åº“æœªè¿æ¥")
        print(f"   æ‰§è¡ŒæŸ¥è¯¢: {query}")
        return f"æŸ¥è¯¢ç»“æœ: æ¨¡æ‹Ÿæ•°æ®"
    
    def get_connection_info(self):
        return {
            "connection_id": self.connection_id,
            "is_connected": self.is_connected,
            "connection_string": getattr(self, 'connection_string', None)
        }

class Logger(metaclass=SingletonMeta):
    """æ—¥å¿—å•ä¾‹ç±»"""
    
    def __init__(self):
        self.logs = []
        self.logger_id = id(self)
        print(f"   åˆ›å»ºæ—¥å¿—å®ä¾‹: {self.logger_id}")
    
    def log(self, message, level="INFO"):
        log_entry = f"[{level}] {message}"
        self.logs.append(log_entry)
        print(f"   æ—¥å¿—è®°å½•: {log_entry}")
    
    def get_logs(self):
        return self.logs.copy()
    
    def clear_logs(self):
        self.logs.clear()
        print(f"   æ¸…ç©ºæ—¥å¿—")

# æµ‹è¯•å•ä¾‹æ¨¡å¼
print(f"   å•ä¾‹æ¨¡å¼æµ‹è¯•:")

# æ•°æ®åº“è¿æ¥å•ä¾‹æµ‹è¯•
print(f"\n     æ•°æ®åº“è¿æ¥å•ä¾‹æµ‹è¯•:")
db1 = DatabaseConnection()
db2 = DatabaseConnection()

print(f"\n     å®ä¾‹æ¯”è¾ƒ:")
print(f"       db1 id: {id(db1)}")
print(f"       db2 id: {id(db2)}")
print(f"       db1 is db2: {db1 is db2}")

db1.connect("localhost:5432/mydb")
db2.connect("localhost:5432/otherdb")  # ä¸ä¼šé‡æ–°è¿æ¥

result1 = db1.execute_query("SELECT * FROM users")
result2 = db2.execute_query("SELECT * FROM products")

print(f"\n     è¿æ¥ä¿¡æ¯:")
print(f"       db1: {db1.get_connection_info()}")
print(f"       db2: {db2.get_connection_info()}")

# æ—¥å¿—å•ä¾‹æµ‹è¯•
print(f"\n     æ—¥å¿—å•ä¾‹æµ‹è¯•:")
logger1 = Logger()
logger2 = Logger()

print(f"\n     å®ä¾‹æ¯”è¾ƒ:")
print(f"       logger1 id: {id(logger1)}")
print(f"       logger2 id: {id(logger2)}")
print(f"       logger1 is logger2: {logger1 is logger2}")

logger1.log("ç³»ç»Ÿå¯åŠ¨", "INFO")
logger2.log("ç”¨æˆ·ç™»å½•", "INFO")
logger1.log("é”™è¯¯å‘ç”Ÿ", "ERROR")

print(f"\n     æ—¥å¿—è®°å½•:")
for log in logger1.get_logs():
    print(f"       {log}")

# 2. å»ºé€ è€…æ¨¡å¼ (Builder Pattern)
print(f"\nğŸ“ 2. å»ºé€ è€…æ¨¡å¼ (Builder Pattern):")

class Computer:
    """è®¡ç®—æœºç±»"""
    
    def __init__(self):
        self.cpu = None
        self.memory = None
        self.storage = None
        self.gpu = None
        self.os = None
        self.price = 0
    
    def __str__(self):
        return f"è®¡ç®—æœºé…ç½®:\n" \
               f"  CPU: {self.cpu}\n" \
               f"  å†…å­˜: {self.memory}\n" \
               f"  å­˜å‚¨: {self.storage}\n" \
               f"  æ˜¾å¡: {self.gpu}\n" \
               f"  æ“ä½œç³»ç»Ÿ: {self.os}\n" \
               f"  ä»·æ ¼: Â¥{self.price}"

class ComputerBuilder:
    """è®¡ç®—æœºå»ºé€ è€…"""
    
    def __init__(self):
        self.computer = Computer()
    
    def set_cpu(self, cpu):
        self.computer.cpu = cpu
        return self
    
    def set_memory(self, memory):
        self.computer.memory = memory
        return self
    
    def set_storage(self, storage):
        self.computer.storage = storage
        return self
    
    def set_gpu(self, gpu):
        self.computer.gpu = gpu
        return self
    
    def set_os(self, os):
        self.computer.os = os
        return self
    
    def calculate_price(self):
        price = 0
        if self.computer.cpu:
            price += 2000
        if self.computer.memory:
            price += 800
        if self.computer.storage:
            price += 500
        if self.computer.gpu:
            price += 3000
        if self.computer.os:
            price += 1000
        self.computer.price = price
        return self
    
    def build(self):
        self.calculate_price()
        return self.computer

class GamingComputerBuilder(ComputerBuilder):
    """æ¸¸æˆç”µè„‘å»ºé€ è€…"""
    
    def build_gaming_computer(self):
        return (self.set_cpu("Intel i9-12900K")
                .set_memory("32GB DDR4")
                .set_storage("1TB NVMe SSD")
                .set_gpu("RTX 4080")
                .set_os("Windows 11")
                .build())

class OfficeComputerBuilder(ComputerBuilder):
    """åŠå…¬ç”µè„‘å»ºé€ è€…"""
    
    def build_office_computer(self):
        return (self.set_cpu("Intel i5-12400")
                .set_memory("16GB DDR4")
                .set_storage("512GB SSD")
                .set_gpu("é›†æˆæ˜¾å¡")
                .set_os("Windows 10")
                .build())

class ComputerDirector:
    """è®¡ç®—æœºæŒ‡æŒ¥è€…"""
    
    def __init__(self, builder):
        self.builder = builder
    
    def construct_computer(self):
        return self.builder.build()
    
    def change_builder(self, builder):
        self.builder = builder

# æµ‹è¯•å»ºé€ è€…æ¨¡å¼
print(f"\n   å»ºé€ è€…æ¨¡å¼æµ‹è¯•:")

# æ‰‹åŠ¨å»ºé€ 
print(f"\n     æ‰‹åŠ¨å»ºé€ è®¡ç®—æœº:")
custom_computer = (ComputerBuilder()
                    .set_cpu("AMD Ryzen 7 5800X")
                    .set_memory("16GB DDR4")
                    .set_storage("1TB SSD")
                    .set_gpu("RTX 3060")
                    .set_os("Ubuntu 20.04")
                    .build())

print(f"       {custom_computer}")

# ä½¿ç”¨ä¸“é—¨å»ºé€ è€…
print(f"\n     ä½¿ç”¨ä¸“é—¨å»ºé€ è€…:")
gaming_builder = GamingComputerBuilder()
gaming_computer = gaming_builder.build_gaming_computer()
print(f"       {gaming_computer}")

office_builder = OfficeComputerBuilder()
office_computer = office_builder.build_office_computer()
print(f"\n       {office_computer}")

# ä½¿ç”¨æŒ‡æŒ¥è€…
print(f"\n     ä½¿ç”¨æŒ‡æŒ¥è€…:")
director = ComputerDirector(gaming_builder)
director_computer = director.construct_computer()
print(f"       {director_computer}")

director.change_builder(office_builder)
director_computer2 = director.construct_computer()
print(f"\n       {director_computer2}")

# 3. åŸå‹æ¨¡å¼ (Prototype Pattern)
print(f"\nğŸ“ 3. åŸå‹æ¨¡å¼ (Prototype Pattern):")

import copy

class Document:
    """æ–‡æ¡£ç±» - åŸå‹"""
    
    def __init__(self, title, content, author):
        self.title = title
        self.content = content
        self.author = author
        self.created_at = None
        self.modified_at = None
        import datetime
        self.created_at = datetime.datetime.now()
        self.modified_at = self.created_at
    
    def clone(self):
        """å…‹éš†æ–‡æ¡£"""
        return copy.deepcopy(self)
    
    def modify_content(self, new_content):
        self.content = new_content
        import datetime
        self.modified_at = datetime.datetime.now()
    
    def __str__(self):
        return f"æ–‡æ¡£: {self.title}\n" \
               f"  ä½œè€…: {self.author}\n" \
               f"  å†…å®¹: {self.content}\n" \
               f"  åˆ›å»ºæ—¶é—´: {self.created_at}\n" \
               f"  ä¿®æ”¹æ—¶é—´: {self.modified_at}"

class DocumentTemplate:
    """æ–‡æ¡£æ¨¡æ¿ç±»"""
    
    def __init__(self):
        self.templates = {}
    
    def add_template(self, name, document):
        """æ·»åŠ æ¨¡æ¿"""
        self.templates[name] = document
        print(f"   æ·»åŠ æ¨¡æ¿: {name}")
    
    def create_document(self, template_name, title, author):
        """åŸºäºæ¨¡æ¿åˆ›å»ºæ–‡æ¡£"""
        if template_name not in self.templates:
            raise ValueError(f"æ¨¡æ¿ {template_name} ä¸å­˜åœ¨")
        
        template = self.templates[template_name]
        document = template.clone()
        document.title = title
        document.author = author
        
        print(f"   åŸºäºæ¨¡æ¿ {template_name} åˆ›å»ºæ–‡æ¡£: {title}")
        return document
    
    def get_templates(self):
        return list(self.templates.keys())

# æµ‹è¯•åŸå‹æ¨¡å¼
print(f"\n   åŸå‹æ¨¡å¼æµ‹è¯•:")

# åˆ›å»ºæ¨¡æ¿
report_template = Document(
    "æŠ¥å‘Šæ¨¡æ¿",
    "1. å¼•è¨€\n2. åˆ†æ\n3. ç»“è®º",
    "ç³»ç»Ÿ"
)

letter_template = Document(
    "ä¿¡ä»¶æ¨¡æ¿",
    "å°Šæ•¬çš„[æ”¶ä»¶äºº]:\n\n[å†…å®¹]\n\næ­¤è‡´\næ•¬ç¤¼",
    "å‘ä»¶äºº"
)

# åˆ›å»ºæ¨¡æ¿ç®¡ç†å™¨
template_manager = DocumentTemplate()
template_manager.add_template("report", report_template)
template_manager.add_template("letter", letter_template)

print(f"\n     å¯ç”¨æ¨¡æ¿: {template_manager.get_templates()}")

# åŸºäºæ¨¡æ¿åˆ›å»ºæ–‡æ¡£
print(f"\n     åŸºäºæ¨¡æ¿åˆ›å»ºæ–‡æ¡£:")
report1 = template_manager.create_document("report", "2023å¹´åº¦æŠ¥å‘Š", "å¼ ä¸‰")
report2 = template_manager.create_document("report", "2024å¹´åº¦æŠ¥å‘Š", "æå››")

letter1 = template_manager.create_document("letter", "é‚€è¯·å‡½", "ç‹äº”")

print(f"\n     åŸå§‹æ¨¡æ¿:")
print(f"       {report_template}")

print(f"\n     åˆ›å»ºçš„æ–‡æ¡£:")
print(f"       {report1}")

# ä¿®æ”¹æ–‡æ¡£
print(f"\n     ä¿®æ”¹æ–‡æ¡£:")
report1.modify_content("1. 2023å¹´æ¦‚è¿°\n2. è´¢åŠ¡åˆ†æ\n3. æœªæ¥å±•æœ›")
print(f"       {report1}")

# éªŒè¯ç‹¬ç«‹æ€§
print(f"\n     éªŒè¯æ–‡æ¡£ç‹¬ç«‹æ€§:")
print(f"       report1ä¿®æ”¹åï¼Œreport2æ˜¯å¦å—å½±å“: {report2.content == report_template.content}")

print(f"\nâœ… åˆ›å»ºå‹è®¾è®¡æ¨¡å¼å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å•ä¾‹ã€å»ºé€ è€…ã€åŸå‹æ¨¡å¼çš„å®ç°")
print(f"   âœ“ ç†è§£åˆ›å»ºå‹æ¨¡å¼çš„åº”ç”¨åœºæ™¯")
print(f"   âœ“ èƒ½é€‰æ‹©åˆé€‚çš„åˆ›å»ºå‹æ¨¡å¼è§£å†³é—®é¢˜")
````

````python
print("ğŸ”§ ç»“æ„å‹è®¾è®¡æ¨¡å¼:")
print("=" * 50)

# 1. é€‚é…å™¨æ¨¡å¼ (Adapter Pattern)
print(f"ğŸ“ 1. é€‚é…å™¨æ¨¡å¼ (Adapter Pattern):")

class MediaPlayer:
    """åª’ä½“æ’­æ”¾å™¨æ¥å£"""
    
    def play(self, audio_type, filename):
        raise NotImplementedError

class AdvancedMediaPlayer:
    """é«˜çº§åª’ä½“æ’­æ”¾å™¨æ¥å£"""
    
    def play_vlc(self, filename):
        raise NotImplementedError
    
    def play_mp4(self, filename):
        raise NotImplementedError

class VlcPlayer(AdvancedMediaPlayer):
    """VLCæ’­æ”¾å™¨"""
    
    def play_vlc(self, filename):
        print(f"   æ’­æ”¾VLCæ–‡ä»¶: {filename}")
    
    def play_mp4(self, filename):
        pass

class Mp4Player(AdvancedMediaPlayer):
    """MP4æ’­æ”¾å™¨"""
    
    def play_vlc(self, filename):
        pass
    
    def play_mp4(self, filename):
        print(f"   æ’­æ”¾MP4æ–‡ä»¶: {filename}")

class MediaAdapter(MediaPlayer):
    """åª’ä½“é€‚é…å™¨"""
    
    def __init__(self, audio_type):
        if audio_type == "vlc":
            self.advanced_player = VlcPlayer()
        elif audio_type == "mp4":
            self.advanced_player = Mp4Player()
        else:
            raise ValueError(f"ä¸æ”¯æŒçš„éŸ³é¢‘ç±»å‹: {audio_type}")
    
    def play(self, audio_type, filename):
        if audio_type == "vlc":
            self.advanced_player.play_vlc(filename)
        elif audio_type == "mp4":
            self.advanced_player.play_mp4(filename)

class AudioPlayer(MediaPlayer):
    """éŸ³é¢‘æ’­æ”¾å™¨"""
    
    def __init__(self):
        self.adapters = {}
    
    def play(self, audio_type, filename):
        if audio_type in ["mp3", "wav"]:
            print(f"   æ’­æ”¾{audio_type.upper()}æ–‡ä»¶: {filename}")
        elif audio_type in ["vlc", "mp4"]:
            if audio_type not in self.adapters:
                self.adapters[audio_type] = MediaAdapter(audio_type)
            self.adapters[audio_type].play(audio_type, filename)
        else:
            print(f"   ä¸æ”¯æŒçš„éŸ³é¢‘æ ¼å¼: {audio_type}")

# æµ‹è¯•é€‚é…å™¨æ¨¡å¼
print(f"   é€‚é…å™¨æ¨¡å¼æµ‹è¯•:")

player = AudioPlayer()

print(f"\n     æ’­æ”¾ä¸åŒæ ¼å¼çš„éŸ³é¢‘:")
player.play("mp3", "song.mp3")
player.play("wav", "music.wav")
player.play("vlc", "movie.vlc")
player.play("mp4", "video.mp4")
player.play("avi", "video.avi")

# 2. è£…é¥°å™¨æ¨¡å¼ (Decorator Pattern)
print(f"\nğŸ“ 2. è£…é¥°å™¨æ¨¡å¼ (Decorator Pattern):")

class Notifier:
    """é€šçŸ¥å™¨æ¥å£"""
    
    def send(self, message):
        raise NotImplementedError

class EmailNotifier(Notifier):
    """é‚®ä»¶é€šçŸ¥å™¨"""
    
    def send(self, message):
        print(f"   å‘é€é‚®ä»¶: {message}")
        return "é‚®ä»¶å‘é€æˆåŠŸ"

class SMSNotifier(Notifier):
    """çŸ­ä¿¡é€šçŸ¥å™¨"""
    
    def send(self, message):
        print(f"   å‘é€çŸ­ä¿¡: {message}")
        return "çŸ­ä¿¡å‘é€æˆåŠŸ"

class NotifierDecorator(Notifier):
    """é€šçŸ¥å™¨è£…é¥°å™¨åŸºç±»"""
    
    def __init__(self, notifier):
        self._notifier = notifier
    
    def send(self, message):
        return self._notifier.send(message)

class EncryptionDecorator(NotifierDecorator):
    """åŠ å¯†è£…é¥°å™¨"""
    
    def send(self, message):
        encrypted_message = self._encrypt(message)
        result = self._notifier.send(encrypted_message)
        print(f"   æ¶ˆæ¯å·²åŠ å¯†")
        return result
    
    def _encrypt(self, message):
        # ç®€å•çš„åŠ å¯†æ¨¡æ‹Ÿ
        return f"[ENCRYPTED]{message}[ENCRYPTED]"

class LoggingDecorator(NotifierDecorator):
    """æ—¥å¿—è£…é¥°å™¨"""
    
    def __init__(self, notifier):
        super().__init__(notifier)
        self._log = []
    
    def send(self, message):
        import datetime
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log_entry = f"[{timestamp}] å‘é€æ¶ˆæ¯: {message}"
        self._log.append(log_entry)
        print(f"   è®°å½•æ—¥å¿—: {log_entry}")
        
        result = self._notifier.send(message)
        return result
    
    def get_log(self):
        return self._log.copy()

class RetryDecorator(NotifierDecorator):
    """é‡è¯•è£…é¥°å™¨"""
    
    def __init__(self, notifier, max_retries=3):
        super().__init__(notifier)
        self.max_retries = max_retries
    
    def send(self, message):
        for attempt in range(self.max_retries):
            try:
                result = self._notifier.send(message)
                if attempt > 0:
                    print(f"   ç¬¬{attempt + 1}æ¬¡å°è¯•æˆåŠŸ")
                return result
            except Exception as e:
                print(f"   ç¬¬{attempt + 1}æ¬¡å°è¯•å¤±è´¥: {e}")
                if attempt == self.max_retries - 1:
                    print(f"   è¾¾åˆ°æœ€å¤§é‡è¯•æ¬¡æ•°ï¼Œå‘é€å¤±è´¥")
                    raise
        return None

# æµ‹è¯•è£…é¥°å™¨æ¨¡å¼
print(f"\n   è£…é¥°å™¨æ¨¡å¼æµ‹è¯•:")

# åŸºç¡€é€šçŸ¥å™¨
print(f"\n     åŸºç¡€é€šçŸ¥å™¨:")
email_notifier = EmailNotifier()
email_notifier.send("é‡è¦é€šçŸ¥")

# æ·»åŠ è£…é¥°å™¨
print(f"\n     æ·»åŠ æ—¥å¿—è£…é¥°å™¨:")
logged_email = LoggingDecorator(email_notifier)
logged_email.send("å¸¦æ—¥å¿—çš„é€šçŸ¥")

print(f"\n     æ·»åŠ åŠ å¯†è£…é¥°å™¨:")
encrypted_email = EncryptionDecorator(email_notifier)
encrypted_email.send("åŠ å¯†é€šçŸ¥")

# ç»„åˆè£…é¥°å™¨
print(f"\n     ç»„åˆå¤šä¸ªè£…é¥°å™¨:")
full_featured_email = RetryDecorator(
    LoggingDecorator(
        EncryptionDecorator(email_notifier)
    ),
    max_retries=2
)
full_featured_email.send("å®Œæ•´åŠŸèƒ½é€šçŸ¥")

print(f"\n     æ—¥å¿—è®°å½•:")
for log in logged_email.get_log():
    print(f"       {log}")

# 3. ä»£ç†æ¨¡å¼ (Proxy Pattern)
print(f"\nğŸ“ 3. ä»£ç†æ¨¡å¼ (Proxy Pattern):")

class Image:
    """å›¾åƒæ¥å£"""
    
    def display(self):
        raise NotImplementedError

class RealImage(Image):
    """çœŸå®å›¾åƒç±»"""
    
    def __init__(self, filename):
        self.filename = filename
        self._load_image_from_disk()
    
    def _load_image_from_disk(self):
        print(f"   ä»ç£ç›˜åŠ è½½å›¾åƒ: {self.filename}")
        # æ¨¡æ‹ŸåŠ è½½æ—¶é—´
        import time
        time.sleep(0.1)
    
    def display(self):
        print(f"   æ˜¾ç¤ºå›¾åƒ: {self.filename}")

class ProxyImage(Image):
    """å›¾åƒä»£ç†ç±»"""
    
    def __init__(self, filename):
        self.filename = filename
        self._real_image = None
        self._access_count = 0
    
    def display(self):
        self._access_count += 1
        print(f"   è®¿é—®æ¬¡æ•°: {self._access_count}")
        
        if self._real_image is None:
            print(f"   é¦–æ¬¡è®¿é—®ï¼Œåˆ›å»ºçœŸå®å›¾åƒå¯¹è±¡")
            self._real_image = RealImage(self.filename)
        
        self._real_image.display()
    
    def get_access_count(self):
        return self._access_count

class DatabaseProxy:
    """æ•°æ®åº“ä»£ç†"""
    
    def __init__(self, connection_string):
        self.connection_string = connection_string
        self._connection = None
        self._query_cache = {}
        self._access_log = []
    
    def _connect(self):
        if self._connection is None:
            print(f"   å»ºç«‹æ•°æ®åº“è¿æ¥: {self.connection_string}")
            self._connection = f"æ¨¡æ‹Ÿè¿æ¥åˆ° {self.connection_string}"
        return self._connection
    
    def execute_query(self, query):
        import datetime
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        
        # è®°å½•è®¿é—®æ—¥å¿—
        self._access_log.append({
            "timestamp": timestamp,
            "query": query
        })
        
        # æ£€æŸ¥ç¼“å­˜
        if query in self._query_cache:
            print(f"   ä»ç¼“å­˜è¿”å›æŸ¥è¯¢ç»“æœ: {query}")
            return self._query_cache[query]
        
        # æ‰§è¡ŒæŸ¥è¯¢
        connection = self._connect()
        print(f"   æ‰§è¡ŒæŸ¥è¯¢: {query}")
        result = f"æŸ¥è¯¢ç»“æœ: {query} çš„æ•°æ®"
        
        # ç¼“å­˜ç»“æœ
        self._query_cache[query] = result
        return result
    
    def get_access_log(self):
        return self._access_log.copy()
    
    def clear_cache(self):
        self._query_cache.clear()
        print(f"   æ¸…ç©ºæŸ¥è¯¢ç¼“å­˜")
    
    def disconnect(self):
        if self._connection is not None:
            print(f"   æ–­å¼€æ•°æ®åº“è¿æ¥")
            self._connection = None

# æµ‹è¯•ä»£ç†æ¨¡å¼
print(f"\n   ä»£ç†æ¨¡å¼æµ‹è¯•:")

# å›¾åƒä»£ç†æµ‹è¯•
print(f"\n     å›¾åƒä»£ç†æµ‹è¯•:")
proxy_image = ProxyImage("photo.jpg")

print(f"\n     é¦–æ¬¡æ˜¾ç¤º:")
proxy_image.display()

print(f"\n     å†æ¬¡æ˜¾ç¤º:")
proxy_image.display()

print(f"\n     ç¬¬ä¸‰æ¬¡æ˜¾ç¤º:")
proxy_image.display()

print(f"\n     æ€»è®¿é—®æ¬¡æ•°: {proxy_image.get_access_count()}")

# æ•°æ®åº“ä»£ç†æµ‹è¯•
print(f"\n     æ•°æ®åº“ä»£ç†æµ‹è¯•:")
db_proxy = DatabaseProxy("localhost:5432/mydb")

print(f"\n     æ‰§è¡ŒæŸ¥è¯¢:")
result1 = db_proxy.execute_query("SELECT * FROM users")
print(f"       {result1}")

result2 = db_proxy.execute_query("SELECT * FROM products")
print(f"       {result2}")

result3 = db_proxy.execute_query("SELECT * FROM users")  # ä»ç¼“å­˜è¿”å›
print(f"       {result3}")

print(f"\n     è®¿é—®æ—¥å¿—:")
for log in db_proxy.get_access_log():
    print(f"       {log['timestamp']}: {log['query']}")

print(f"\nâœ… ç»“æ„å‹è®¾è®¡æ¨¡å¼å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡é€‚é…å™¨ã€è£…é¥°å™¨ã€ä»£ç†æ¨¡å¼çš„å®ç°")
print(f"   âœ“ ç†è§£ç»“æ„å‹æ¨¡å¼çš„è®¾è®¡åŸç†å’Œåº”ç”¨åœºæ™¯")
print(f"   âœ“ èƒ½é€‰æ‹©åˆé€‚çš„ç»“æ„å‹æ¨¡å¼è§£å†³é—®é¢˜")
````

````python
print("ğŸ¯ è¡Œä¸ºå‹è®¾è®¡æ¨¡å¼:")
print("=" * 50)

# 1. è´£ä»»é“¾æ¨¡å¼ (Chain of Responsibility Pattern)
print(f"ğŸ“ 1. è´£ä»»é“¾æ¨¡å¼ (Chain of Responsibility Pattern):")

class Handler:
    """å¤„ç†å™¨æŠ½è±¡åŸºç±»"""
    
    def __init__(self):
        self._next_handler = None
    
    def set_next(self, handler):
        """è®¾ç½®ä¸‹ä¸€ä¸ªå¤„ç†å™¨"""
        self._next_handler = handler
        return handler
    
    def handle(self, request):
        """å¤„ç†è¯·æ±‚"""
        result = self._process_request(request)
        
        if not result and self._next_handler:
            result = self._next_handler.handle(request)
        
        return result
    
    def _process_request(self, request):
        """å¤„ç†è¯·æ±‚çš„å…·ä½“é€»è¾‘"""
        raise NotImplementedError

class Level1Support(Handler):
    """ä¸€çº§æ”¯æŒ"""
    
    def _process_request(self, request):
        if request['level'] == 1:
            print(f"   ä¸€çº§æ”¯æŒå¤„ç†: {request['issue']}")
            return "ä¸€çº§æ”¯æŒå·²è§£å†³é—®é¢˜"
        return None

class Level2Support(Handler):
    """äºŒçº§æ”¯æŒ"""
    
    def _process_request(self, request):
        if request['level'] == 2:
            print(f"   äºŒçº§æ”¯æŒå¤„ç†: {request['issue']}")
            return "äºŒçº§æ”¯æŒå·²è§£å†³é—®é¢˜"
        return None

class Level3Support(Handler):
    """ä¸‰çº§æ”¯æŒ"""
    
    def _process_request(self, request):
        if request['level'] == 3:
            print(f"   ä¸‰çº§æ”¯æŒå¤„ç†: {request['issue']}")
            return "ä¸‰çº§æ”¯æŒå·²è§£å†³é—®é¢˜"
        return None

class ManagerSupport(Handler):
    """ç»ç†æ”¯æŒ"""
    
    def _process_request(self, request):
        print(f"   ç»ç†å¤„ç†ç´§æ€¥é—®é¢˜: {request['issue']}")
        return "ç»ç†å·²å¤„ç†é—®é¢˜"

# æµ‹è¯•è´£ä»»é“¾æ¨¡å¼
print(f"   è´£ä»»é“¾æ¨¡å¼æµ‹è¯•:")

# åˆ›å»ºè´£ä»»é“¾
level1 = Level1Support()
level2 = Level2Support()
level3 = Level3Support()
manager = ManagerSupport()

# è®¾ç½®è´£ä»»é“¾
level1.set_next(level2).set_next(level3).set_next(manager)

# æµ‹è¯•ä¸åŒçº§åˆ«çš„è¯·æ±‚
requests = [
    {'level': 1, 'issue': 'å¯†ç é‡ç½®'},
    {'level': 2, 'issue': 'è½¯ä»¶å®‰è£…é—®é¢˜'},
    {'level': 3, 'issue': 'ç³»ç»Ÿå´©æºƒ'},
    {'level': 4, 'issue': 'ç´§æ€¥æ•…éšœ'}
]

print(f"\n     å¤„ç†å®¢æˆ·è¯·æ±‚:")
for request in requests:
    print(f"\n       å¤„ç†è¯·æ±‚: {request}")
    result = level1.handle(request)
    print(f"         ç»“æœ: {result}")

# 2. çŠ¶æ€æ¨¡å¼ (State Pattern)
print(f"\nğŸ“ 2. çŠ¶æ€æ¨¡å¼ (State Pattern):")

class OrderState:
    """è®¢å•çŠ¶æ€æŠ½è±¡åŸºç±»"""
    
    def handle(self, order):
        raise NotImplementedError

class PendingState(OrderState):
    """å¾…å¤„ç†çŠ¶æ€"""
    
    def handle(self, order):
        print(f"   è®¢å• {order.order_id} æ­£åœ¨å¾…å¤„ç†")
        order.set_state(ProcessingState())
        return "è®¢å•å¼€å§‹å¤„ç†"

class ProcessingState(OrderState):
    """å¤„ç†ä¸­çŠ¶æ€"""
    
    def handle(self, order):
        print(f"   è®¢å• {order.order_id} æ­£åœ¨å¤„ç†ä¸­")
        order.set_state(ShippedState())
        return "è®¢å•å·²å‘è´§"

class ShippedState(OrderState):
    """å·²å‘è´§çŠ¶æ€"""
    
    def handle(self, order):
        print(f"   è®¢å• {order.order_id} å·²å‘è´§")
        order.set_state(DeliveredState())
        return "è®¢å•å·²é€è¾¾"

class DeliveredState(OrderState):
    """å·²é€è¾¾çŠ¶æ€"""
    
    def handle(self, order):
        print(f"   è®¢å• {order.order_id} å·²é€è¾¾")
        return "è®¢å•å·²å®Œæˆ"

class CancelledState(OrderState):
    """å·²å–æ¶ˆçŠ¶æ€"""
    
    def handle(self, order):
        print(f"   è®¢å• {order.order_id} å·²å–æ¶ˆ")
        return "è®¢å•å·²å–æ¶ˆ"

class Order:
    """è®¢å•ç±»"""
    
    def __init__(self, order_id):
        self.order_id = order_id
        self._state = PendingState()
        self.history = []
    
    def set_state(self, state):
        self._state = state
        self.history.append(state.__class__.__name__)
    
    def process(self):
        return self._state.handle(self)
    
    def cancel(self):
        if isinstance(self._state, (PendingState, ProcessingState)):
            self.set_state(CancelledState())
            return "è®¢å•å·²å–æ¶ˆ"
        return "è®¢å•æ— æ³•å–æ¶ˆ"
    
    def get_state(self):
        return self._state.__class__.__name__
    
    def get_history(self):
        return self.history.copy()

# æµ‹è¯•çŠ¶æ€æ¨¡å¼
print(f"\n   çŠ¶æ€æ¨¡å¼æµ‹è¯•:")

# åˆ›å»ºè®¢å•
order = Order("ORD001")
print(f"\n     è®¢å•åˆå§‹çŠ¶æ€: {order.get_state()}")

# å¤„ç†è®¢å•
print(f"\n     è®¢å•å¤„ç†æµç¨‹:")
result1 = order.process()
print(f"       {result1} - å½“å‰çŠ¶æ€: {order.get_state()}")

result2 = order.process()
print(f"       {result2} - å½“å‰çŠ¶æ€: {order.get_state()}")

result3 = order.process()
print(f"       {result3} - å½“å‰çŠ¶æ€: {order.get_state()}")

result4 = order.process()
print(f"       {result4} - å½“å‰çŠ¶æ€: {order.get_state()}")

print(f"\n     çŠ¶æ€å˜æ›´å†å²: {order.get_history()}")

# æµ‹è¯•è®¢å•å–æ¶ˆ
print(f"\n     æµ‹è¯•è®¢å•å–æ¶ˆ:")
order2 = Order("ORD002")
order2.process()  # è¿›å…¥å¤„ç†ä¸­çŠ¶æ€
cancel_result = order2.cancel()
print(f"       {cancel_result} - å½“å‰çŠ¶æ€: {order2.get_state()}")

# 3. æ¨¡æ¿æ–¹æ³•æ¨¡å¼ (Template Method Pattern)
print(f"\nğŸ“ 3. æ¨¡æ¿æ–¹æ³•æ¨¡å¼ (Template Method Pattern):")

from abc import ABC, abstractmethod

class DataProcessor(ABC):
    """æ•°æ®å¤„ç†å™¨æŠ½è±¡åŸºç±»"""
    
    def process_data(self, data):
        """æ¨¡æ¿æ–¹æ³• - å®šä¹‰å¤„ç†æµç¨‹"""
        print(f"   å¼€å§‹å¤„ç†æ•°æ®...")
        
        validated_data = self.validate_data(data)
        print(f"   æ•°æ®éªŒè¯å®Œæˆ")
        
        transformed_data = self.transform_data(validated_data)
        print(f"   æ•°æ®è½¬æ¢å®Œæˆ")
        
        result = self.analyze_data(transformed_data)
        print(f"   æ•°æ®åˆ†æå®Œæˆ")
        
        self.save_result(result)
        print(f"   ç»“æœä¿å­˜å®Œæˆ")
        
        return result
    
    @abstractmethod
    def validate_data(self, data):
        """éªŒè¯æ•°æ® - å­ç±»å®ç°"""
        pass
    
    @abstractmethod
    def transform_data(self, data):
        """è½¬æ¢æ•°æ® - å­ç±»å®ç°"""
        pass
    
    @abstractmethod
    def analyze_data(self, data):
        """åˆ†ææ•°æ® - å­ç±»å®ç°"""
        pass
    
    def save_result(self, result):
        """ä¿å­˜ç»“æœ - é»˜è®¤å®ç°ï¼Œå­ç±»å¯é‡å†™"""
        print(f"   ä¿å­˜ç»“æœ: {result}")

class SalesDataProcessor(DataProcessor):
    """é”€å”®æ•°æ®å¤„ç†å™¨"""
    
    def validate_data(self, data):
        # éªŒè¯é”€å”®æ•°æ®
        if not isinstance(data, list):
            raise ValueError("é”€å”®æ•°æ®å¿…é¡»æ˜¯åˆ—è¡¨")
        
        validated = []
        for item in data:
            if isinstance(item, dict) and 'amount' in item and 'date' in item:
                if item['amount'] > 0:
                    validated.append(item)
        
        print(f"   éªŒè¯é€šè¿‡ {len(validated)}/{len(data)} æ¡é”€å”®è®°å½•")
        return validated
    
    def transform_data(self, data):
        # è½¬æ¢é”€å”®æ•°æ®
        transformed = []
        for item in data:
            transformed_item = {
                'amount': float(item['amount']),
                'date': item['date'],
                'month': item['date'][:7]  # æå–æœˆä»½
            }
            transformed.append(transformed_item)
        
        print(f"   è½¬æ¢äº† {len(transformed)} æ¡è®°å½•")
        return transformed
    
    def analyze_data(self, data):
        # åˆ†æé”€å”®æ•°æ®
        total_sales = sum(item['amount'] for item in data)
        avg_sales = total_sales / len(data) if data else 0
        
        # æŒ‰æœˆä»½ç»Ÿè®¡
        monthly_sales = {}
        for item in data:
            month = item['month']
            monthly_sales[month] = monthly_sales.get(month, 0) + item['amount']
        
        result = {
            'total_sales': total_sales,
            'average_sales': avg_sales,
            'monthly_sales': monthly_sales,
            'record_count': len(data)
        }
        
        print(f"   åˆ†æç»“æœ: æ€»é”€å”®é¢ {total_sales}, å¹³å‡é”€å”®é¢ {avg_sales:.2f}")
        return result
    
    def save_result(self, result):
        print(f"   ä¿å­˜é”€å”®åˆ†ææŠ¥å‘Šåˆ°æ•°æ®åº“")
        print(f"   æŠ¥å‘Šæ‘˜è¦: æ€»é”€å”®é¢ Â¥{result['total_sales']}, è®°å½•æ•° {result['record_count']}")

class UserDataProcessor(DataProcessor):
    """ç”¨æˆ·æ•°æ®å¤„ç†å™¨"""
    
    def validate_data(self, data):
        # éªŒè¯ç”¨æˆ·æ•°æ®
        if not isinstance(data, list):
            raise ValueError("ç”¨æˆ·æ•°æ®å¿…é¡»æ˜¯åˆ—è¡¨")
        
        validated = []
        for item in data:
            if isinstance(item, dict) and 'name' in item and 'age' in item:
                if isinstance(item['age'], int) and 0 < item['age'] < 150:
                    validated.append(item)
        
        print(f"   éªŒè¯é€šè¿‡ {len(validated)}/{len(data)} æ¡ç”¨æˆ·è®°å½•")
        return validated
    
    def transform_data(self, data):
        # è½¬æ¢ç”¨æˆ·æ•°æ®
        transformed = []
        for item in data:
            transformed_item = {
                'name': item['name'].strip(),
                'age': item['age'],
                'age_group': self._get_age_group(item['age'])
            }
            transformed.append(transformed_item)
        
        print(f"   è½¬æ¢äº† {len(transformed)} æ¡è®°å½•")
        return transformed
    
    def _get_age_group(self, age):
        if age < 18:
            return "æœªæˆå¹´"
        elif age < 30:
            return "é’å¹´"
        elif age < 50:
            return "ä¸­å¹´"
        else:
            return "è€å¹´"
    
    def analyze_data(self, data):
        # åˆ†æç”¨æˆ·æ•°æ®
        total_users = len(data)
        avg_age = sum(item['age'] for item in data) / total_users if data else 0
        
        # æŒ‰å¹´é¾„ç»„ç»Ÿè®¡
        age_groups = {}
        for item in data:
            group = item['age_group']
            age_groups[group] = age_groups.get(group, 0) + 1
        
        result = {
            'total_users': total_users,
            'average_age': avg_age,
            'age_distribution': age_groups
        }
        
        print(f"   åˆ†æç»“æœ: æ€»ç”¨æˆ·æ•° {total_users}, å¹³å‡å¹´é¾„ {avg_age:.1f}")
        return result
    
    def save_result(self, result):
        print(f"   ä¿å­˜ç”¨æˆ·åˆ†ææŠ¥å‘Šåˆ°æ–‡ä»¶")
        print(f"   æŠ¥å‘Šæ‘˜è¦: æ€»ç”¨æˆ·æ•° {result['total_users']}, å¹³å‡å¹´é¾„ {result['average_age']:.1f}")

# æµ‹è¯•æ¨¡æ¿æ–¹æ³•æ¨¡å¼
print(f"\n   æ¨¡æ¿æ–¹æ³•æ¨¡å¼æµ‹è¯•:")

# é”€å”®æ•°æ®å¤„ç†
print(f"\n     é”€å”®æ•°æ®å¤„ç†:")
sales_data = [
    {'amount': 100, 'date': '2023-01-01'},
    {'amount': 200, 'date': '2023-01-15'},
    {'amount': 150, 'date': '2023-02-01'},
    {'amount': -50, 'date': '2023-02-15'},  # æ— æ•ˆæ•°æ®
    {'amount': 300, 'date': '2023-03-01'}
]

sales_processor = SalesDataProcessor()
sales_result = sales_processor.process_data(sales_data)

# ç”¨æˆ·æ•°æ®å¤„ç†
print(f"\n     ç”¨æˆ·æ•°æ®å¤„ç†:")
user_data = [
    {'name': 'å¼ ä¸‰', 'age': 25},
    {'name': 'æå››', 'age': 30},
    {'name': 'ç‹äº”', 'age': 17},
    {'name': 'èµµå…­', 'age': 55},
    {'name': 'é’±ä¸ƒ', 'age': 200}  # æ— æ•ˆæ•°æ®
]

user_processor = UserDataProcessor()
user_result = user_processor.process_data(user_data)

print(f"\nâœ… è¡Œä¸ºå‹è®¾è®¡æ¨¡å¼å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡è´£ä»»é“¾ã€çŠ¶æ€ã€æ¨¡æ¿æ–¹æ³•æ¨¡å¼çš„å®ç°")
print(f"   âœ“ ç†è§£è¡Œä¸ºå‹æ¨¡å¼çš„è®¾è®¡åŸç†å’Œåº”ç”¨åœºæ™¯")
print(f"   âœ“ èƒ½é€‰æ‹©åˆé€‚çš„è¡Œä¸ºå‹æ¨¡å¼è§£å†³é—®é¢˜")
````
