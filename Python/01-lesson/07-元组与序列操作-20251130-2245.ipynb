### åºåˆ—å®é™…åº”ç”¨ [â­â­è¿›é˜¶]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šé€šè¿‡å®é™…åº”ç”¨æ¡ˆä¾‹å±•ç¤ºå…ƒç»„å’Œåºåˆ—åœ¨çœŸå®åœºæ™¯ä¸­çš„ä½¿ç”¨æ–¹æ³•ã€‚

- æŒæ¡åºåˆ—åœ¨å®é™…é—®é¢˜ä¸­çš„åº”ç”¨
- ç†è§£åºåˆ—æ•°æ®å¤„ç†çš„æ€è·¯

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°å®Œæ•´çš„åºåˆ—åº”ç”¨

- ä¼˜åŒ–åºåˆ—æ“ä½œæ•ˆç‡

```python
print("ğŸ¯ åºåˆ—å®é™…åº”ç”¨:")
print("=" * 50)

# æ¡ˆä¾‹1ï¼šåæ ‡ç³»ç»Ÿå’Œå‡ ä½•è®¡ç®—
print(f"ğŸ“ 1. åæ ‡ç³»ç»Ÿå’Œå‡ ä½•è®¡ç®—:")

class Point2D:
    """äºŒç»´ç‚¹ç±»"""
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __repr__(self):
        return f"Point2D({self.x}, {self.y})"
    
    def __add__(self, other):
        return Point2D(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Point2D(self.x - other.x, self.y - other.y)
    
    def distance_to(self, other):
        """è®¡ç®—åˆ°å¦ä¸€ä¸ªç‚¹çš„è·ç¦»"""
        dx = self.x - other.x
        dy = self.y - other.y
        return (dx**2 + dy**2) ** 0.5
    
    def to_tuple(self):
        """è½¬æ¢ä¸ºå…ƒç»„"""
        return (self.x, self.y)

# å‡ ä½•è®¡ç®—å‡½æ•°
def calculate_triangle_area(p1, p2, p3):
    """è®¡ç®—ä¸‰è§’å½¢é¢ç§¯"""
    # ä½¿ç”¨æµ·ä¼¦å…¬å¼
    a = p1.distance_to(p2)
    b = p2.distance_to(p3)
    c = p3.distance_to(p1)
    
    s = (a + b + c) / 2
    area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
    return area

def find_closest_point(origin, points):
    """æ‰¾åˆ°ç¦»åŸç‚¹æœ€è¿‘çš„ç‚¹"""
    if not points:
        return None
    
    closest = points[0]
    min_distance = origin.distance_to(closest)
    
    for point in points[1:]:
        distance = origin.distance_to(point)
        if distance < min_distance:
            min_distance = distance
            closest = point
    
    return closest, min_distance

# æµ‹è¯•å‡ ä½•è®¡ç®—
origin = Point2D(0, 0)
points = [
    Point2D(3, 4),
    Point2D(1, 1),
    Point2D(5, 12),
    Point2D(-2, -2),
    Point2D(8, 6)
]

print(f"   åŸç‚¹: {origin}")
print(f"   ç‚¹é›†åˆ: {points}")

# è®¡ç®—è·ç¦»
print(f"\n   è·ç¦»è®¡ç®—:")
for i, point in enumerate(points):
    distance = origin.distance_to(point)
    print(f"     {point} åˆ°åŸç‚¹è·ç¦»: {distance:.2f}")

# æ‰¾æœ€è¿‘ç‚¹
closest_point, min_dist = find_closest_point(origin, points)
print(f"\n   æœ€è¿‘çš„ç‚¹: {closest_point}, è·ç¦»: {min_dist:.2f}")

# è®¡ç®—ä¸‰è§’å½¢é¢ç§¯
if len(points) >= 3:
    triangle_area = calculate_triangle_area(points[0], points[1], points[2])
    print(f"\n   ä¸‰è§’å½¢é¢ç§¯: {triangle_area:.2f}")

# æ¡ˆä¾‹2ï¼šæ—¶é—´åºåˆ—æ•°æ®å¤„ç†
print(f"\nğŸ“ 2. æ—¶é—´åºåˆ—æ•°æ®å¤„ç†:")

def process_time_series(data):
    """å¤„ç†æ—¶é—´åºåˆ—æ•°æ®"""
    print(f"   åŸå§‹æ•°æ®ç‚¹æ•°: {len(data)}")
    
    # æ•°æ®æ¸…æ´— - ç§»é™¤å¼‚å¸¸å€¼
    if data:
        mean_value = sum(data) / len(data)
        std_dev = (sum((x - mean_value) ** 2 for x in data) / len(data)) ** 0.5
        
        # ç§»é™¤è¶…å‡º3å€æ ‡å‡†å·®çš„æ•°æ®
        cleaned_data = [x for x in data if abs(x - mean_value) <= 3 * std_dev]
        print(f"   æ¸…æ´—åæ•°æ®ç‚¹æ•°: {len(cleaned_data)}")
    else:
        cleaned_data = data
    
    # è®¡ç®—ç»Ÿè®¡æŒ‡æ ‡
    if cleaned_data:
        stats = {
            'mean': sum(cleaned_data) / len(cleaned_data),
            'min': min(cleaned_data),
            'max': max(cleaned_data),
            'count': len(cleaned_data)
        }
        
        # è®¡ç®—ç§»åŠ¨å¹³å‡
        window_size = 3
        moving_avg = []
        for i in range(len(cleaned_data) - window_size + 1):
            window = cleaned_data[i:i + window_size]
            avg = sum(window) / window_size
            moving_avg.append(avg)
        
        # è®¡ç®—è¶‹åŠ¿ï¼ˆç®€å•çº¿æ€§å›å½’æ–œç‡ï¼‰
        n = len(cleaned_data)
        if n > 1:
            x_mean = (n - 1) / 2
            y_mean = stats['mean']
            
            numerator = sum((i - x_mean) * (cleaned_data[i] - y_mean) for i in range(n))
            denominator = sum((i - x_mean) ** 2 for i in range(n))
            
            slope = numerator / denominator if denominator != 0 else 0
        else:
            slope = 0
        
        return {
            'cleaned_data': cleaned_data,
            'statistics': stats,
            'moving_average': moving_avg,
            'trend_slope': slope
        }
    
    return None

# æµ‹è¯•æ—¶é—´åºåˆ—å¤„ç†
time_series_data = [10, 12, 15, 14, 16, 18, 20, 19, 22, 25, 24, 26, 28, 30, 100, 29, 31]
print(f"   æ—¶é—´åºåˆ—æ•°æ®: {time_series_data}")

result = process_time_series(time_series_data)

if result:
    print(f"\n   å¤„ç†ç»“æœ:")
    stats = result['statistics']
    print(f"     ç»Ÿè®¡ä¿¡æ¯:")
    print(f"       å¹³å‡å€¼: {stats['mean']:.2f}")
    print(f"       æœ€å°å€¼: {stats['min']}")
    print(f"       æœ€å¤§å€¼: {stats['max']}")
    print(f"       æ•°æ®ç‚¹æ•°: {stats['count']}")
    
    print(f"\n     ç§»åŠ¨å¹³å‡(çª—å£3): {result['moving_average'][:5]}...")
    print(f"     è¶‹åŠ¿æ–œç‡: {result['trend_slope']:.4f}")
    
    if result['trend_slope'] > 0:
        print(f"     è¶‹åŠ¿: ä¸Šå‡")
    elif result['trend_slope'] < 0:
        print(f"     è¶‹åŠ¿: ä¸‹é™")
    else:
        print(f"     è¶‹åŠ¿: å¹³ç¨³")

# æ¡ˆä¾‹3ï¼šæ•°æ®å‹ç¼©å’Œç¼–ç 
print(f"\nğŸ“ 3. æ•°æ®å‹ç¼©å’Œç¼–ç :")

def run_length_encoding(data):
    """è¡Œç¨‹é•¿åº¦ç¼–ç """
    if not data:
        return []
    
    encoded = []
    current_char = data[0]
    count = 1
    
    for char in data[1:]:
        if char == current_char:
            count += 1
        else:
            encoded.append((current_char, count))
            current_char = char
            count = 1
    
    encoded.append((current_char, count))
    return encoded

def run_length_decoding(encoded_data):
    """è¡Œç¨‹é•¿åº¦è§£ç """
    decoded = []
    for char, count in encoded_data:
        decoded.extend([char] * count)
    return decoded

def calculate_compression_ratio(original, compressed):
    """è®¡ç®—å‹ç¼©æ¯”"""
    original_size = len(original)
    compressed_size = len(compressed) * 2  # æ¯ä¸ªç¼–ç é¡¹åŒ…å«å­—ç¬¦å’Œè®¡æ•°
    ratio = compressed_size / original_size if original_size > 0 else 0
    return ratio

# æµ‹è¯•æ•°æ®å‹ç¼©
test_strings = [
    "AAAABBBCCDAA",
    "Hello World",
    "1111222333444",
    "PythonPythonPython"
]

for test_str in test_strings:
    print(f"\n   æµ‹è¯•å­—ç¬¦ä¸²: '{test_str}'")
    
    # ç¼–ç 
    encoded = run_length_encoding(test_str)
    print(f"     ç¼–ç ç»“æœ: {encoded}")
    
    # è§£ç 
    decoded = ''.join(run_length_decoding(encoded))
    print(f"     è§£ç ç»“æœ: '{decoded}'")
    
    # éªŒè¯
    is_correct = decoded == test_str
    print(f"     è§£ç æ­£ç¡®: {is_correct}")
    
    # å‹ç¼©æ¯”
    ratio = calculate_compression_ratio(test_str, encoded)
    print(f"     å‹ç¼©æ¯”: {ratio:.2f}")

# æ¡ˆä¾‹4ï¼šæ•°æ®èšåˆå’Œåˆ†æ
print(f"\nğŸ“ 4. æ•°æ®èšåˆå’Œåˆ†æ:")

def analyze_sales_data(sales_records):
    """åˆ†æé”€å”®æ•°æ®"""
    print(f"   é”€å”®è®°å½•æ•°: {len(sales_records)}")
    
    # æŒ‰äº§å“åˆ†ç»„
    product_sales = {}
    for record in sales_records:
        product = record[0]
        amount = record[1]
        
        if product not in product_sales:
            product_sales[product] = []
        product_sales[product].append(amount)
    
    # è®¡ç®—æ¯ä¸ªäº§å“çš„ç»Ÿè®¡ä¿¡æ¯
    product_stats = {}
    for product, amounts in product_sales.items():
        product_stats[product] = {
            'total': sum(amounts),
            'count': len(amounts),
            'average': sum(amounts) / len(amounts),
            'min': min(amounts),
            'max': max(amounts)
        }
    
    # æ‰¾å‡ºæœ€ä½³å’Œæœ€å·®è¡¨ç°
    best_product = max(product_stats.items(), key=lambda x: x[1]['total'])
    worst_product = min(product_stats.items(), key=lambda x: x[1]['total'])
    
    # è®¡ç®—æ€»ä½“ç»Ÿè®¡
    all_amounts = [record[1] for record in sales_records]
    total_sales = sum(all_amounts)
    average_sale = total_sales / len(all_amounts)
    
    return {
        'product_stats': product_stats,
        'best_product': best_product,
        'worst_product': worst_product,
        'total_sales': total_sales,
        'average_sale': average_sale
    }

# æµ‹è¯•é”€å”®æ•°æ®åˆ†æ
sales_data = [
    ("ç¬”è®°æœ¬ç”µè„‘", 5999),
    ("æ— çº¿é¼ æ ‡", 99),
    ("ç¬”è®°æœ¬ç”µè„‘", 5999),
    ("æœºæ¢°é”®ç›˜", 299),
    ("æ— çº¿é¼ æ ‡", 99),
    ("æ˜¾ç¤ºå™¨", 1999),
    ("ç¬”è®°æœ¬ç”µè„‘", 5999),
    ("æœºæ¢°é”®ç›˜", 299),
    ("æ— çº¿é¼ æ ‡", 99),
    ("æ˜¾ç¤ºå™¨", 1999)
]

print(f"   é”€å”®æ•°æ®: {sales_data}")

analysis = analyze_sales_data(sales_data)

print(f"\n   åˆ†æç»“æœ:")
print(f"     æ€»é”€å”®é¢: Â¥{analysis['total_sales']}")
print(f"     å¹³å‡é”€å”®é¢: Â¥{analysis['average_sale']:.2f}")

print(f"\n   å„äº§å“ç»Ÿè®¡:")
for product, stats in analysis['product_stats'].items():
    print(f"     {product}:")
    print(f"       é”€å”®é¢: Â¥{stats['total']}")
    print(f"       é”€å”®é‡: {stats['count']}")
    print(f"       å¹³å‡ä»·: Â¥{stats['average']:.2f}")

print(f"\n   æœ€ä½³äº§å“: {analysis['best_product'][0]} (Â¥{analysis['best_product'][1]['total']})")
print(f"   æœ€å·®äº§å“: {analysis['worst_product'][0]} (Â¥{analysis['worst_product'][1]['total']})")

# æ¡ˆä¾‹5ï¼šæ•°æ®è½¬æ¢å’Œæ˜ å°„
print(f"\nğŸ“ 5. æ•°æ®è½¬æ¢å’Œæ˜ å°„:")

def data_transformation_pipeline(data):
    """æ•°æ®è½¬æ¢ç®¡é“"""
    print(f"   è¾“å…¥æ•°æ®: {data}")
    
    # æ­¥éª¤1: æ•°æ®æ ‡å‡†åŒ–
    normalized = tuple(str(item).strip().lower() for item in data if item is not None)
    print(f"   æ ‡å‡†åŒ–å: {normalized}")
    
    # æ­¥éª¤2: æ•°æ®è¿‡æ»¤
    filtered = tuple(item for item in normalized if len(item) > 2)
    print(f"   è¿‡æ»¤å(é•¿åº¦>2): {filtered}")
    
    # æ­¥éª¤3: æ•°æ®æ˜ å°„
    mapping = {
        'apple': 'è‹¹æœ',
        'banana': 'é¦™è•‰',
        'orange': 'æ©™å­',
        'grape': 'è‘¡è„'
    }
    
    mapped = tuple(mapping.get(item, item) for item in filtered)
    print(f"   æ˜ å°„å: {mapped}")
    
    # æ­¥éª¤4: æ•°æ®èšåˆ
    if mapped:
        unique_items = tuple(sorted(set(mapped)))
        item_counts = tuple((item, mapped.count(item)) for item in unique_items)
        print(f"   èšåˆç»“æœ: {item_counts}")
        
        return {
            'original': data,
            'normalized': normalized,
            'filtered': filtered,
            'mapped': mapped,
            'aggregated': item_counts
        }
    
    return None

# æµ‹è¯•æ•°æ®è½¬æ¢
raw_data = ["Apple", " banana ", "Orange", "grape", "APPLE", None, "Banana", "pear"]
transform_result = data_transformation_pipeline(raw_data)

if transform_result:
    print(f"\n   è½¬æ¢ç®¡é“å®Œæˆï¼Œæœ€ç»ˆç»“æœ: {transform_result['aggregated']}")
```
