````python
print("ğŸ› ï¸ å¸¸ç”¨è£…é¥°å™¨æ¨¡å¼:")
print("=" * 50)

# 1. æ—¥å¿—è£…é¥°å™¨
print(f"ğŸ“ 1. æ—¥å¿—è£…é¥°å™¨:")

import time
import functools

def log_decorator(level="INFO"):
    """æ—¥å¿—è£…é¥°å™¨å·¥å‚"""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
            print(f"   [{level}] {timestamp} - è°ƒç”¨å‡½æ•°: {func.__name__}")
            print(f"   [{level}] {timestamp} - å‚æ•°: args={args}, kwargs={kwargs}")
            
            start_time = time.time()
            try:
                result = func(*args, **kwargs)
                end_time = time.time()
                print(f"   [{level}] {timestamp} - å‡½æ•°æ‰§è¡ŒæˆåŠŸ")
                print(f"   [{level}] {timestamp} - æ‰§è¡Œæ—¶é—´: {end_time - start_time:.4f}ç§’")
                print(f"   [{level}] {timestamp} - è¿”å›å€¼: {result}")
                return result
            except Exception as e:
                end_time = time.time()
                print(f"   [{ERROR}] {timestamp} - å‡½æ•°æ‰§è¡Œå¤±è´¥: {str(e)}")
                print(f"   [{ERROR}] {timestamp} - æ‰§è¡Œæ—¶é—´: {end_time - start_time:.4f}ç§’")
                raise
        return wrapper
    return decorator

# ä½¿ç”¨æ—¥å¿—è£…é¥°å™¨
@log_decorator("DEBUG")
def process_data(data, multiplier=2):
    """å¤„ç†æ•°æ®"""
    time.sleep(0.1)  # æ¨¡æ‹Ÿå¤„ç†æ—¶é—´
    return [x * multiplier for x in data if isinstance(x, (int, float))]

print(f"   æ—¥å¿—è£…é¥°å™¨æµ‹è¯•:")
result = process_data([1, 2, 3, 4, 5], multiplier=3)
print(f"   æœ€ç»ˆç»“æœ: {result}")

# 2. ç¼“å­˜è£…é¥°å™¨
print(f"\nğŸ“ 2. ç¼“å­˜è£…é¥°å™¨:")

def cache_decorator(max_size=128):
    """LRUç¼“å­˜è£…é¥°å™¨"""
    def decorator(func):
        cache = {}
        access_order = []
        
        @functools.wraps(func)
        def wrapper(*args):
            # åˆ›å»ºç¼“å­˜é”®
            key = str(args)
            
            # æ£€æŸ¥ç¼“å­˜
            if key in cache:
                # æ›´æ–°è®¿é—®é¡ºåº
                access_order.remove(key)
                access_order.append(key)
                print(f"   ç¼“å­˜å‘½ä¸­: {key}")
                return cache[key]
            
            # è®¡ç®—æ–°å€¼
            print(f"   è®¡ç®—æ–°å€¼: {key}")
            result = func(*args)
            
            # ç¼“å­˜ç®¡ç†
            if len(cache) >= max_size:
                # ç§»é™¤æœ€ä¹…æœªè®¿é—®çš„é¡¹
                oldest_key = access_order.pop(0)
                del cache[oldest_key]
                print(f"   ç¼“å­˜å·²æ»¡ï¼Œç§»é™¤: {oldest_key}")
            
            # æ·»åŠ æ–°é¡¹
            cache[key] = result
            access_order.append(key)
            
            return result
        
        def cache_info():
            return {
                'size': len(cache),
                'max_size': max_size,
                'keys': list(cache.keys())
            }
        
        def cache_clear():
            cache.clear()
            access_order.clear()
        
        wrapper.cache_info = cache_info
        wrapper.cache_clear = cache_clear
        return wrapper
    return decorator

@cache_decorator(max_size=3)
def fibonacci(n):
    """è®¡ç®—æ–æ³¢é‚£å¥‘æ•°ï¼ˆé€’å½’ç‰ˆæœ¬ï¼‰"""
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(f"   ç¼“å­˜è£…é¥°å™¨æµ‹è¯•:")
print(f"     fibonacci(5): {fibonacci(5)}")
print(f"     fibonacci(6): {fibonacci(6)}")
print(f"     fibonacci(5): {fibonacci(5)}")  # ä»ç¼“å­˜è·å–
print(f"     ç¼“å­˜ä¿¡æ¯: {fibonacci.cache_info()}")

# 3. é‡è¯•è£…é¥°å™¨
print(f"\nğŸ“ 3. é‡è¯•è£…é¥°å™¨:")

def retry_decorator(max_attempts=3, delay=1, exceptions=(Exception,)):
    """é‡è¯•è£…é¥°å™¨"""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(max_attempts):
                try:
                    return func(*args, **kwargs)
                except exceptions as e:
                    if attempt == max_attempts - 1:
                        print(f"   é‡è¯•å¤±è´¥ï¼Œå·²è¾¾åˆ°æœ€å¤§å°è¯•æ¬¡æ•°: {max_attempts}")
                        raise
                    else:
                        print(f"   ç¬¬{attempt + 1}æ¬¡å°è¯•å¤±è´¥: {str(e)}ï¼Œ{delay}ç§’åé‡è¯•...")
                        time.sleep(delay)
        return wrapper
    return decorator

# æ¨¡æ‹Ÿä¸ç¨³å®šçš„å‡½æ•°
import random

@retry_decorator(max_attempts=3, delay=0.1, exceptions=(ValueError,))
def unstable_function():
    """ä¸ç¨³å®šçš„å‡½æ•°ï¼Œæœ‰ä¸€å®šæ¦‚ç‡å¤±è´¥"""
    if random.random() < 0.7:  # 70%æ¦‚ç‡å¤±è´¥
        raise ValueError("éšæœºå¤±è´¥")
    return "æˆåŠŸ!"

print(f"   é‡è¯•è£…é¥°å™¨æµ‹è¯•:")
try:
    result = unstable_function()
    print(f"   ç»“æœ: {result}")
except ValueError as e:
    print(f"   æœ€ç»ˆå¤±è´¥: {e}")

# 4. æ€§èƒ½ç›‘æ§è£…é¥°å™¨
print(f"\nğŸ“ 4. æ€§èƒ½ç›‘æ§è£…é¥°å™¨:")

def performance_monitor(threshold=1.0):
    """æ€§èƒ½ç›‘æ§è£…é¥°å™¨"""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            start_time = time.time()
            start_memory = 0  # ç®€åŒ–ç‰ˆæœ¬ï¼Œå®é™…å¯ä»¥ä½¿ç”¨psutilç­‰åº“
            
            try:
                result = func(*args, **kwargs)
                end_time = time.time()
                execution_time = end_time - start_time
                
                # æ€§èƒ½è­¦å‘Š
                if execution_time > threshold:
                    print(f"   âš ï¸  æ€§èƒ½è­¦å‘Š: {func.__name__} æ‰§è¡Œæ—¶é—´ {execution_time:.4f}ç§’ è¶…è¿‡é˜ˆå€¼ {threshold}ç§’")
                else:
                    print(f"   âœ… æ€§èƒ½æ­£å¸¸: {func.__name__} æ‰§è¡Œæ—¶é—´ {execution_time:.4f}ç§’")
                
                return result
            
            except Exception as e:
                end_time = time.time()
                execution_time = end_time - start_time
                print(f"   âŒ æ‰§è¡Œå¤±è´¥: {func.__name__} åœ¨ {execution_time:.4f}ç§’åå¤±è´¥: {str(e)}")
                raise
        
        return wrapper
    return decorator

@performance_monitor(threshold=0.5)
def fast_operation():
    """å¿«é€Ÿæ“ä½œ"""
    time.sleep(0.1)
    return "å¿«é€Ÿå®Œæˆ"

@performance_monitor(threshold=0.5)
def slow_operation():
    """æ…¢é€Ÿæ“ä½œ"""
    time.sleep(1.0)
    return "æ…¢é€Ÿå®Œæˆ"

print(f"   æ€§èƒ½ç›‘æ§æµ‹è¯•:")
fast_result = fast_operation()
slow_result = slow_operation()

# 5. å‚æ•°éªŒè¯è£…é¥°å™¨
print(f"\nğŸ“ 5. å‚æ•°éªŒè¯è£…é¥°å™¨:")

def validate_types(**type_hints):
    """å‚æ•°ç±»å‹éªŒè¯è£…é¥°å™¨"""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            # è·å–å‡½æ•°å‚æ•°å
            import inspect
            sig = inspect.signature(func)
            bound_args = sig.bind(*args, **kwargs)
            bound_args.apply_defaults()
            
            # éªŒè¯å‚æ•°ç±»å‹
            for param_name, param_value in bound_args.arguments.items():
                if param_name in type_hints:
                    expected_type = type_hints[param_name]
                    if not isinstance(param_value, expected_type):
                        raise TypeError(
                            f"å‚æ•° '{param_name}' æœŸæœ›ç±»å‹ {expected_type.__name__}, "
                            f"å®é™…ç±»å‹ {type(param_value).__name__}"
                        )
            
            return func(*args, **kwargs)
        return wrapper
    return decorator

def validate_range(**ranges):
    """å‚æ•°èŒƒå›´éªŒè¯è£…é¥°å™¨"""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            import inspect
            sig = inspect.signature(func)
            bound_args = sig.bind(*args, **kwargs)
            bound_args.apply_defaults()
            
            for param_name, param_value in bound_args.arguments.items():
                if param_name in ranges:
                    min_val, max_val = ranges[param_name]
                    if not (min_val <= param_value <= max_val):
                        raise ValueError(
                            f"å‚æ•° '{param_name}' å€¼ {param_value} è¶…å‡ºèŒƒå›´ [{min_val}, {max_val}]"
                        )
            
            return func(*args, **kwargs)
        return wrapper
    return decorator

@validate_types(name=str, age=int, score=float)
@validate_range(age=(0, 150), score=(0.0, 100.0))
def create_student(name, age, score):
    """åˆ›å»ºå­¦ç”Ÿä¿¡æ¯"""
    return {"name": name, "age": age, "score": score}

print(f"   å‚æ•°éªŒè¯è£…é¥°å™¨æµ‹è¯•:")
try:
    student = create_student("å¼ ä¸‰", 20, 85.5)
    print(f"   æˆåŠŸåˆ›å»º: {student}")
    
    # æµ‹è¯•ç±»å‹é”™è¯¯
    # invalid_student = create_student("æå››", "25", 90.0)
    
    # æµ‹è¯•èŒƒå›´é”™è¯¯
    # invalid_student = create_student("ç‹äº”", 200, 85.5)
    
except (TypeError, ValueError) as e:
    print(f"   éªŒè¯å¤±è´¥: {e}")

print(f"\nâœ… å¸¸ç”¨è£…é¥°å™¨æ¨¡å¼å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å„ç§è£…é¥°å™¨æ¨¡å¼")
print(f"   âœ“ ç†è§£è£…é¥°å™¨æœ€ä½³å®è·µ")
print(f"   âœ“ èƒ½åº”ç”¨è£…é¥°å™¨è§£å†³å®é™…é—®é¢˜")
````

````python
print("ğŸš€ é«˜çº§è£…é¥°å™¨æŠ€æœ¯:")
print("=" * 50)

# 1. ç±»è£…é¥°å™¨
print(f"ğŸ“ 1. ç±»è£…é¥°å™¨:")

class CountCalls:
    """è®¡æ•°è°ƒç”¨è£…é¥°å™¨ç±»"""
    def __init__(self, func):
        self.func = func
        self.count = 0
        functools.update_wrapper(self, func)
    
    def __call__(self, *args, **kwargs):
        self.count += 1
        print(f"   å‡½æ•° {self.func.__name__} å·²è¢«è°ƒç”¨ {self.count} æ¬¡")
        return self.func(*args, **kwargs)
    
    def get_count(self):
        return self.count
    
    def reset_count(self):
        self.count = 0

@CountCalls
def greet_user(name):
    return f"Hello, {name}!"

print(f"   ç±»è£…é¥°å™¨æµ‹è¯•:")
print(f"     {greet_user('å¼ ä¸‰')}")
print(f"     {greet_user('æå››')}")
print(f"     {greet_user('ç‹äº”')}")
print(f"     æ€»è°ƒç”¨æ¬¡æ•°: {greet_user.get_count()}")

greet_user.reset_count()
print(f"     é‡ç½®åè°ƒç”¨æ¬¡æ•°: {greet_user.get_count()}")

# 2. è£…é¥°å™¨é“¾
print(f"\nğŸ“ 2. è£…é¥°å™¨é“¾:")

def decorator1(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"   è£…é¥°å™¨1 - å¼€å§‹")
        result = func(*args, **kwargs)
        print(f"   è£…é¥°å™¨1 - ç»“æŸ")
        return result
    return wrapper

def decorator2(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"   è£…é¥°å™¨2 - å¼€å§‹")
        result = func(*args, **kwargs)
        print(f"   è£…é¥°å™¨2 - ç»“æŸ")
        return result
    return wrapper

def decorator3(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"   è£…é¥°å™¨3 - å¼€å§‹")
        result = func(*args, **kwargs)
        print(f"   è£…é¥°å™¨3 - ç»“æŸ")
        return result
    return wrapper

@decorator1
@decorator2
@decorator3
def chained_function():
    print(f"   åŸå‡½æ•°æ‰§è¡Œ")
    return "é“¾å¼è£…é¥°å®Œæˆ"

print(f"   è£…é¥°å™¨é“¾æµ‹è¯•:")
result = chained_function()
print(f"   æœ€ç»ˆç»“æœ: {result}")

# 3. å¸¦çŠ¶æ€è£…é¥°å™¨
print(f"\nğŸ“ 3. å¸¦çŠ¶æ€è£…é¥°å™¨:")

def stateful_decorator(initial_state=None):
    """å¸¦çŠ¶æ€çš„è£…é¥°å™¨å·¥å‚"""
    def decorator(func):
        state = initial_state or {}
        
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            # è®¿é—®å’Œä¿®æ”¹çŠ¶æ€
            wrapper.call_count = state.get('call_count', 0) + 1
            state['call_count'] = wrapper.call_count
            
            wrapper.last_args = args
            wrapper.last_kwargs = kwargs
            
            print(f"   ç¬¬{wrapper.call_count}æ¬¡è°ƒç”¨")
            result = func(*args, **kwargs)
            wrapper.last_result = result
            
            return result
        
        # åˆå§‹åŒ–çŠ¶æ€å±æ€§
        wrapper.call_count = 0
        wrapper.last_args = None
        wrapper.last_kwargs = None
        wrapper.last_result = None
        wrapper.get_state = lambda: state.copy()
        
        return wrapper
    return decorator

@stateful_decorator({'call_count': 0})
def calculate(a, b, operation='add'):
    """è®¡ç®—å‡½æ•°"""
    if operation == 'add':
        return a + b
    elif operation == 'multiply':
        return a * b
    else:
        return 0

print(f"   å¸¦çŠ¶æ€è£…é¥°å™¨æµ‹è¯•:")
result1 = calculate(5, 3)
result2 = calculate(10, 2, 'multiply')
result3 = calculate(7, 8)

print(f"\n   çŠ¶æ€ä¿¡æ¯:")
print(f"     è°ƒç”¨æ¬¡æ•°: {calculate.call_count}")
print(f"     æœ€åå‚æ•°: {calculate.last_args}, {calculate.last_kwargs}")
print(f"     æœ€åç»“æœ: {calculate.last_result}")
print(f"     å®Œæ•´çŠ¶æ€: {calculate.get_state()}")

# 4. è£…é¥°å™¨å·¥å‚çš„é«˜çº§ç”¨æ³•
print(f"\nğŸ“ 4. è£…é¥°å™¨å·¥å‚çš„é«˜çº§ç”¨æ³•:")

def advanced_decorator_factory(**config):
    """é«˜çº§è£…é¥°å™¨å·¥å‚"""
    def decorator(func):
        # ä»é…ç½®ä¸­è·å–å‚æ•°
        log_calls = config.get('log_calls', False)
        cache_results = config.get('cache_results', False)
        validate_args = config.get('validate_args', False)
        
        cache = {} if cache_results else None
        
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            # æ—¥å¿—è®°å½•
            if log_calls:
                print(f"   è°ƒç”¨å‡½æ•°: {func.__name__}({args}, {kwargs})")
            
            # å‚æ•°éªŒè¯
            if validate_args:
                if not args or not isinstance(args[0], (int, float)):
                    raise ValueError("ç¬¬ä¸€ä¸ªå‚æ•°å¿…é¡»æ˜¯æ•°å­—")
            
            # ç¼“å­˜æ£€æŸ¥
            if cache_results:
                key = str(args) + str(sorted(kwargs.items()))
                if key in cache:
                    if log_calls:
                        print(f"   ç¼“å­˜å‘½ä¸­: {key}")
                    return cache[key]
            
            # æ‰§è¡ŒåŸå‡½æ•°
            result = func(*args, **kwargs)
            
            # ç¼“å­˜ç»“æœ
            if cache_results:
                cache[key] = result
                if log_calls:
                    print(f"   ç»“æœå·²ç¼“å­˜: {key}")
            
            # æ—¥å¿—è®°å½•ç»“æœ
            if log_calls:
                print(f"   å‡½æ•°ç»“æœ: {result}")
            
            return result
        
        # æ·»åŠ é¢å¤–æ–¹æ³•
        if cache_results:
            def clear_cache():
                cache.clear()
            def cache_info():
                return {'size': len(cache), 'keys': list(cache.keys())}
            
            wrapper.clear_cache = clear_cache
            wrapper.cache_info = cache_info
        
        return wrapper
    return decorator

# ä½¿ç”¨é«˜çº§è£…é¥°å™¨å·¥å‚
@advanced_decorator_factory(
    log_calls=True,
    cache_results=True,
    validate_args=True
)
def advanced_calculator(x, y, operation='add'):
    """é«˜çº§è®¡ç®—å™¨"""
    if operation == 'add':
        return x + y
    elif operation == 'multiply':
        return x * y
    elif operation == 'power':
        return x ** y
    else:
        return 0

print(f"   é«˜çº§è£…é¥°å™¨å·¥å‚æµ‹è¯•:")
result1 = advanced_calculator(5, 3, 'add')
result2 = advanced_calculator(5, 3, 'add')  # ä»ç¼“å­˜è·å–
result3 = advanced_calculator(2, 8, 'power')

print(f"\n   ç¼“å­˜ä¿¡æ¯: {advanced_calculator.cache_info()}")

# 5. æ–¹æ³•è£…é¥°å™¨
print(f"\nğŸ“ 5. æ–¹æ³•è£…é¥°å™¨:")

def method_decorator(func):
    """æ–¹æ³•è£…é¥°å™¨"""
    @functools.wraps(func)
    def wrapper(self, *args, **kwargs):
        print(f"   æ–¹æ³• {func.__name__} åœ¨ç±» {self.__class__.__name__} ä¸­è¢«è°ƒç”¨")
        print(f"   å®ä¾‹çŠ¶æ€: {getattr(self, '__dict__', {})}")
        result = func(self, *args, **kwargs)
        print(f"   æ–¹æ³•æ‰§è¡Œå®Œæˆï¼Œç»“æœ: {result}")
        return result
    return wrapper

class Calculator:
    def __init__(self, name):
        self.name = name
        self.history = []
    
    @method_decorator
    def add(self, a, b):
        result = a + b
        self.history.append(f"{a} + {b} = {result}")
        return result
    
    @method_decorator
    def multiply(self, a, b):
        result = a * b
        self.history.append(f"{a} * {b} = {result}")
        return result
    
    def get_history(self):
        return self.history.copy()

print(f"   æ–¹æ³•è£…é¥°å™¨æµ‹è¯•:")
calc = Calculator("æˆ‘çš„è®¡ç®—å™¨")
result1 = calc.add(5, 3)
result2 = calc.multiply(4, 6)
print(f"\n   è®¡ç®—å†å²: {calc.get_history()}")

# 6. å±æ€§è£…é¥°å™¨
print(f"\nğŸ“ 6. å±æ€§è£…é¥°å™¨:")

def property_decorator(getter):
    """å±æ€§è£…é¥°å™¨"""
    # åˆ›å»ºå±æ€§å
    prop_name = getter.__name__
    private_name = f'_{prop_name}'
    
    @functools.wraps(getter)
    def wrapper(self):
        if not hasattr(self, private_name):
            # å¦‚æœç§æœ‰å±æ€§ä¸å­˜åœ¨ï¼Œè°ƒç”¨getterè®¡ç®—
            value = getter(self)
            setattr(self, private_name, value)
        return getattr(self, private_name)
    
    return property(wrapper)

class LazyCalculator:
    def __init__(self):
        self._a = 10
        self._b = 20
    
    @property_decorator
    def expensive_result(self):
        """è®¡ç®—å¯†é›†çš„å±æ€§"""
        print(f"   æ­£åœ¨æ‰§è¡Œå¤æ‚è®¡ç®—...")
        time.sleep(0.1)  # æ¨¡æ‹Ÿå¤æ‚è®¡ç®—
        result = sum(range(self._a * self._b))
        print(f"   å¤æ‚è®¡ç®—å®Œæˆ: {result}")
        return result

print(f"   å±æ€§è£…é¥°å™¨æµ‹è¯•:")
lazy_calc = LazyCalculator()
print(f"\n   ç¬¬ä¸€æ¬¡è®¿é—®:")
value1 = lazy_calc.expensive_result
print(f"   ç»“æœ: {value1}")

print(f"\n   ç¬¬äºŒæ¬¡è®¿é—®:")
value2 = lazy_calc.expensive_result
print(f"   ç»“æœ: {value2}")

print(f"\nâœ… é«˜çº§è£…é¥°å™¨æŠ€æœ¯å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡ç±»è£…é¥°å™¨å®ç°åº”ç”¨")
print(f"   âœ“ ç†è§£è£…é¥°å™¨é“¾æ‰§è¡Œæœºåˆ¶")
print(f"   âœ“ èƒ½è®¾è®¡å¤æ‚çŠ¶æ€è£…é¥°å™¨")
````

### å®é™…åº”ç”¨æ¡ˆä¾‹ [â­â­è¿›é˜¶]

````python
print("ğŸ’¼ å®é™…åº”ç”¨æ¡ˆä¾‹:")
print("=" * 50)

# æ¡ˆä¾‹1ï¼šWebæ¡†æ¶è·¯ç”±è£…é¥°å™¨
print(f"ğŸ“ 1. Webæ¡†æ¶è·¯ç”±è£…é¥°å™¨:")

class SimpleWebFramework:
    def __init__(self):
        self.routes = {}
        self.middleware = []
    
    def route(self, path, methods=['GET']):
        """è·¯ç”±è£…é¥°å™¨å·¥å‚"""
        def decorator(func):
            self.routes[path] = {
                'handler': func,
                'methods': methods
            }
            print(f"   æ³¨å†Œè·¯ç”±: {path} -> {func.__name__} ({', '.join(methods)})")
            return func
        return decorator
    
    def use_middleware(self, middleware_func):
        """æ·»åŠ ä¸­é—´ä»¶"""
        self.middleware.append(middleware_func)
        print(f"   æ·»åŠ ä¸­é—´ä»¶: {middleware_func.__name__}")
    
    def handle_request(self, path, method='GET', data=None):
        """å¤„ç†è¯·æ±‚"""
        print(f"\n   å¤„ç†è¯·æ±‚: {method} {path}")
        
        if path not in self.routes:
            return {'error': 'Not Found', 'status': 404}
        
        route = self.routes[path]
        if method not in route['methods']:
            return {'error': 'Method Not Allowed', 'status': 405}
        
        # æ‰§è¡Œä¸­é—´ä»¶é“¾
        request = {'path': path, 'method': method, 'data': data}
        
        for middleware in self.middleware:
            request = middleware(request)
            if 'error' in request:
                return request
        
        # æ‰§è¡Œè·¯ç”±å¤„ç†å™¨
        try:
            result = route['handler'](request)
            return {'data': result, 'status': 200}
        except Exception as e:
            return {'error': str(e), 'status': 500}

# åˆ›å»ºWebæ¡†æ¶å®ä¾‹
app = SimpleWebFramework()

# å®šä¹‰ä¸­é—´ä»¶
def logging_middleware(request):
    print(f"     [æ—¥å¿—] {request['method']} {request['path']}")
    return request

def auth_middleware(request):
    # ç®€åŒ–çš„è®¤è¯æ£€æŸ¥
    if request['path'].startswith('/protected'):
        if not request.get('data', {}).get('authenticated'):
            return {'error': 'Unauthorized', 'status': 401}
    return request

# æ·»åŠ ä¸­é—´ä»¶
app.use_middleware(logging_middleware)
app.use_middleware(auth_middleware)

# å®šä¹‰è·¯ç”±å¤„ç†å™¨
@app.route('/')
def home(request):
    return {'message': 'Welcome to Home Page!'}

@app.route('/users', methods=['GET', 'POST'])
def users(request):
    if request['method'] == 'GET':
        return {'users': ['Alice', 'Bob', 'Charlie']}
    else:
        return {'message': 'User created'}

@app.route('/protected/data')
def protected_data(request):
    return {'secret': 'This is protected data'}

print(f"   Webæ¡†æ¶è£…é¥°å™¨æµ‹è¯•:")

# æµ‹è¯•è¯·æ±‚
response1 = app.handle_request('/')
print(f"     å“åº”: {response1}")

response2 = app.handle_request('/users', 'POST')
print(f"     å“åº”: {response2}")

response3 = app.handle_request('/protected/data')
print(f"     å“åº”: {response3}")

response4 = app.handle_request('/protected/data', 'GET', {'authenticated': True})
print(f"     å“åº”: {response4}")

# æ¡ˆä¾‹2ï¼šAPIéªŒè¯è£…é¥°å™¨ç³»ç»Ÿ
print(f"\nğŸ“ 2. APIéªŒè¯è£…é¥°å™¨ç³»ç»Ÿ:")

def api_validate(**validators):
    """APIå‚æ•°éªŒè¯è£…é¥°å™¨"""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(request):
            data = request.get('data', {})
            errors = {}
            
            # æ‰§è¡ŒéªŒè¯
            for field, validator in validators.items():
                if field in data:
                    try:
                        validator(data[field])
                    except ValueError as e:
                        errors[field] = str(e)
                elif 'required' in validator.__name__.lower():
                    errors[field] = f'{field} is required'
            
            if errors:
                return {'error': 'Validation failed', 'details': errors, 'status': 400}
            
            return func(request)
        return wrapper
    return decorator

# éªŒè¯å™¨å‡½æ•°
def required_string(value):
    if not isinstance(value, str) or not value.strip():
        raise ValueError('Must be a non-empty string')

def email_validator(value):
    required_string(value)
    if '@' not in value:
        raise ValueError('Must be a valid email')

def age_validator(value):
    if not isinstance(value, int) or not (0 <= value <= 150):
        raise ValueError('Must be an integer between 0 and 150')

def optional_string(value):
    if value is not None:
        required_string(value)

# APIç«¯ç‚¹
@api_validate(
    name=required_string,
    email=email_validator,
    age=age_validator,
    phone=optional_string
)
def create_user_api(request):
    data = request['data']
    return {
        'message': 'User created successfully',
        'user': {
            'name': data['name'],
            'email': data['email'],
            'age': data['age'],
            'phone': data.get('phone')
        }
    }

print(f"   APIéªŒè¯è£…é¥°å™¨æµ‹è¯•:")

# æœ‰æ•ˆè¯·æ±‚
valid_request = {
    'data': {
        'name': 'å¼ ä¸‰',
        'email': 'zhangsan@example.com',
        'age': 25,
        'phone': '1234567890'
    }
}

response = create_user_api(valid_request)
print(f"     æœ‰æ•ˆè¯·æ±‚: {response}")

# æ— æ•ˆè¯·æ±‚
invalid_request = {
    'data': {
        'name': '',
        'email': 'invalid-email',
        'age': 200
    }
}

response = create_user_api(invalid_request)
print(f"     æ— æ•ˆè¯·æ±‚: {response}")

# æ¡ˆä¾‹3ï¼šæ•°æ®åº“äº‹åŠ¡è£…é¥°å™¨
print(f"\nğŸ“ 3. æ•°æ®åº“äº‹åŠ¡è£…é¥°å™¨:")

def database_transaction(retry_count=3, isolation_level='READ_COMMITTED'):
    """æ•°æ®åº“äº‹åŠ¡è£…é¥°å™¨"""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(retry_count):
                try:
                    print(f"   å¼€å§‹äº‹åŠ¡ (éš”ç¦»çº§åˆ«: {isolation_level})")
                    print(f"   æ‰§è¡Œæ“ä½œ: {func.__name__}")
                    
                    # æ¨¡æ‹Ÿæ•°æ®åº“æ“ä½œ
                    result = func(*args, **kwargs)
                    
                    print(f"   æ“ä½œæˆåŠŸï¼Œæäº¤äº‹åŠ¡")
                    return result
                
                except Exception as e:
                    print(f"   æ“ä½œå¤±è´¥ï¼Œå›æ»šäº‹åŠ¡: {str(e)}")
                    if attempt == retry_count - 1:
                        print(f"   è¾¾åˆ°æœ€å¤§é‡è¯•æ¬¡æ•°ï¼Œæ”¾å¼ƒæ“ä½œ")
                        raise
                    else:
                        print(f"   å‡†å¤‡é‡è¯• ({attempt + 1}/{retry_count})")
                        time.sleep(0.1)
        return wrapper
    return decorator

# æ¨¡æ‹Ÿæ•°æ®åº“æ“ä½œ
@database_transaction(retry_count=2, isolation_level='SERIALIZABLE')
def update_user_balance(user_id, amount):
    """æ›´æ–°ç”¨æˆ·ä½™é¢"""
    # æ¨¡æ‹Ÿå¯èƒ½çš„æ•°æ®åº“å†²çª
    import random
    if random.random() < 0.5:  # 50%æ¦‚ç‡å¤±è´¥
        raise Exception("æ•°æ®åº“å†²çª")
    
    return f"ç”¨æˆ·{user_id}ä½™é¢æ›´æ–°{amount}å…ƒ"

@database_transaction()
def create_order(user_id, items):
    """åˆ›å»ºè®¢å•"""
    total_amount = sum(item['price'] for item in items)
    return f"ä¸ºç”¨æˆ·{user_id}åˆ›å»ºè®¢å•ï¼Œæ€»é‡‘é¢{total_amount}å…ƒ"

print(f"   æ•°æ®åº“äº‹åŠ¡è£…é¥°å™¨æµ‹è¯•:")

try:
    result1 = update_user_balance(1001, 500)
    print(f"     ç»“æœ: {result1}")
except Exception as e:
    print(f"     æœ€ç»ˆå¤±è´¥: {e}")

result2 = create_order(1002, [
    {'name': 'å•†å“A', 'price': 100},
    {'name': 'å•†å“B', 'price': 200}
])
print(f"     ç»“æœ: {result2}")

# æ¡ˆä¾‹4ï¼šç¼“å­˜å’Œæ€§èƒ½ä¼˜åŒ–è£…é¥°å™¨
print(f"\nğŸ“ 4. ç¼“å­˜å’Œæ€§èƒ½ä¼˜åŒ–è£…é¥°å™¨:")

def smart_cache(ttl=300, max_size=1000, key_func=None):
    """æ™ºèƒ½ç¼“å­˜è£…é¥°å™¨"""
    def decorator(func):
        cache = {}
        timestamps = {}
        access_times = {}
        
        def get_cache_key(*args, **kwargs):
            if key_func:
                return key_func(*args, **kwargs)
            return str(args) + str(sorted(kwargs.items()))
        
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            current_time = time.time()
            cache_key = get_cache_key(*args, **kwargs)
            
            # æ£€æŸ¥ç¼“å­˜
            if cache_key in cache:
                # æ£€æŸ¥TTL
                if current_time - timestamps[cache_key] < ttl:
                    access_times[cache_key] = current_time
                    print(f"   ç¼“å­˜å‘½ä¸­: {cache_key}")
                    return cache[cache_key]
                else:
                    # ç¼“å­˜è¿‡æœŸï¼Œæ¸…ç†
                    del cache[cache_key]
                    del timestamps[cache_key]
                    del access_times[cache_key]
                    print(f"   ç¼“å­˜è¿‡æœŸ: {cache_key}")
            
            # æ£€æŸ¥ç¼“å­˜å¤§å°
            if len(cache) >= max_size:
                # LRUæ¸…ç†
                lru_key = min(access_times.keys(), key=lambda k: access_times[k])
                del cache[lru_key]
                del timestamps[lru_key]
                del access_times[lru_key]
                print(f"   LRUæ¸…ç†: {lru_key}")
            
            # è®¡ç®—æ–°å€¼
            print(f"   è®¡ç®—æ–°å€¼: {cache_key}")
            start_time = time.time()
            result = func(*args, **kwargs)
            execution_time = time.time() - start_time
            
            # ç¼“å­˜ç»“æœ
            cache[cache_key] = result
            timestamps[cache_key] = current_time
            access_times[cache_key] = current_time
            
            print(f"   ç¼“å­˜ä¿å­˜: {cache_key} (æ‰§è¡Œæ—¶é—´: {execution_time:.4f}s)")
            return result
        
        def cache_stats():
            return {
                'size': len(cache),
                'max_size': max_size,
                'ttl': ttl,
                'keys': list(cache.keys())
            }
        
        def clear_cache():
            cache.clear()
            timestamps.clear()
            access_times.clear()
        
        wrapper.cache_stats = cache_stats
        wrapper.clear_cache = clear_cache
        return wrapper
    return decorator

@smart_cache(ttl=5, max_size=3)
def expensive_api_call(user_id, data_type='profile'):
    """æ¨¡æ‹Ÿæ˜‚è´µçš„APIè°ƒç”¨"""
    time.sleep(0.2)  # æ¨¡æ‹Ÿç½‘ç»œå»¶è¿Ÿ
    return {
        'user_id': user_id,
        'data_type': data_type,
        'data': f"æ¨¡æ‹Ÿæ•°æ® for {user_id} - {data_type}",
        'timestamp': time.time()
    }

print(f"   æ™ºèƒ½ç¼“å­˜è£…é¥°å™¨æµ‹è¯•:")

# æµ‹è¯•ç¼“å­˜
result1 = expensive_api_call(1001, 'profile')
result2 = expensive_api_call(1001, 'profile')  # ä»ç¼“å­˜è·å–
result3 = expensive_api_call(1002, 'settings')
result4 = expensive_api_call(1003, 'profile')
result5 = expensive_api_call(1004, 'profile')  # è§¦å‘LRUæ¸…ç†

print(f"\n   ç¼“å­˜ç»Ÿè®¡: {expensive_api_call.cache_stats()}")

# ç­‰å¾…TTLè¿‡æœŸ
print(f"\n   ç­‰å¾…ç¼“å­˜è¿‡æœŸ...")
time.sleep(6)
result6 = expensive_api_call(1001, 'profile')  # ç¼“å­˜å·²è¿‡æœŸ

print(f"\nâœ… å®é™…åº”ç”¨æ¡ˆä¾‹å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡è£…é¥°å™¨å®é™…åº”ç”¨")
print(f"   âœ“ ç†è§£è£…é¥°å™¨æ¶æ„è®¾è®¡")
print(f"   âœ“ èƒ½è®¾è®¡å®Œæ•´è£…é¥°å™¨è§£å†³æ–¹æ¡ˆ")
````
