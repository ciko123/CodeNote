# 03-条件判断语句

```python
# 基础if语句
age = 18
if age >= 18:
    print(f"成年人")

age = 16
if age >= 18:
    print(f"成年人")
else:
    print(f"未成年人")
```

```python
# if-elif-else结构
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f"等级: {grade}")
```

### 三元运算符

```python
# 条件表达式（三元运算符）
temperature = 25
weather = "炎热" if temperature > 30 else "凉爽"
print(f"   温度: {temperature}°C, 天气: {weather}")
```
