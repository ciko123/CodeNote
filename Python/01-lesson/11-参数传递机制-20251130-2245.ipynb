### 2.2 参数传递机制 [⭐基础]
**知识点说明**：Python提供了多种参数传递方式，包括位置参数、关键字参数、默认参数、可变参数等，理解这些机制有助于编写更灵活的函数。

**学习要求**：

- 理解参数传递的顺序规则

**案例要求**：
- 实现不同参数类型的函数
- 进行参数传递机制验证
- 应用参数技巧解决实际问题

````python
print("🔧 位置参数与关键字参数:")
print("=" * 50)

# 位置参数
print(f"📝 1. 位置参数:")
def greet(name, age, city):
    """使用位置参数的问候函数"""
    print(f"你好，我叫{name}，今年{age}岁，来自{city}")

# 位置参数调用（按顺序传递）
print("   位置参数调用:")
greet("张三", 25, "北京")
greet("李四", 30, "上海")

# 关键字参数
print(f"\n📝 2. 关键字参数:")
# 使用参数名指定值，顺序可以改变
print("   关键字参数调用:")
greet(name="王五", age=28, city="广州")
greet(age=32, city="深圳", name="赵六")  # 顺序改变

# 混合使用位置参数和关键字参数
print("\n   混合参数调用:")
greet("孙七", age=35, city="杭州")  # 第一个用位置，后面用关键字
# greet(name="周八", 40, "南京")  # 错误：位置参数必须在关键字参数之前

# 默认参数
print(f"\n📝 3. 默认参数:")
def create_user(username, email, role="user", status="active"):
    """创建用户，角色和状态有默认值"""
    print(f"创建用户: {username}, 邮箱: {email}, 角色: {role}, 状态: {status}")
    return {
        "username": username,
        "email": email,
        "role": role,
        "status": status
    }

# 使用默认参数
print("   使用默认参数:")
user1 = create_user("admin", "admin@example.com")
print(f"   结果: {user1}")

# 覆盖部分默认参数
print("\n   覆盖部分默认参数:")
user2 = create_user("editor", "editor@example.com", role="editor")
print(f"   结果: {user2}")

# 覆盖所有默认参数
print("\n   覆盖所有默认参数:")
user3 = create_user("guest", "guest@example.com", "guest", "inactive")
print(f"   结果: {user3}")

# 使用关键字参数覆盖默认值
print("\n   使用关键字参数:")
user4 = create_user("manager", "manager@example.com", status="busy")
print(f"   结果: {user4}")

print(f"\n✅ 位置参数与关键字参数完成")
print(f"🎯 学习目标达成:")
print(f"   ✓ 掌握位置参数使用")
print(f"   ✓ 理解关键字参数优势")
print(f"   ✓ 能灵活使用默认参数")
````

### 可变参数 [⭐基础]
**知识点说明**：可变参数允许函数接受任意数量的参数，包括*args（位置参数）和**kwargs（关键字参数）。

**学习要求**：
- 掌握*args和**kwargs的使用
- 理解可变参数的应用场景
- 能设计灵活的可变参数函数

**案例要求**：
- 实现可变参数函数
- 进行参数收集和解包练习
- 应用可变参数解决实际问题
- 验证点：能正确使用可变参数

````python
print("🔄 可变参数:")
print("=" * 50)

# *args - 可变位置参数
print(f"📝 1. *args - 可变位置参数:")
def sum_all(*numbers):
    """计算所有数字的和"""
    print(f"   接收到的参数: {numbers}")
    print(f"   参数类型: {type(numbers)}")
    total = sum(numbers)
    print(f"   计算结果: {total}")
    return total

# 调用可变位置参数函数
print("   调用sum_all():")
sum_all(1, 2, 3, 4, 5)
sum_all(10, 20, 30)
sum_all()  # 不传递参数

# **kwargs - 可变关键字参数
print(f"\n📝 2. **kwargs - 可变关键字参数:")
def print_info(**info):
    """打印信息字典"""
    print(f"   接收到的参数: {info}")
    print(f"   参数类型: {type(info)}")
    for key, value in info.items():
        print(f"     {key}: {value}")

# 调用可变关键字参数函数
print("   调用print_info():")
print_info(name="张三", age=25, city="北京")
print_info(product="手机", price=2999, brand="华为", color="黑色")
print_info()  # 不传递参数

# 混合使用*args和**kwargs
print(f"\n📝 3. 混合使用*args和**kwargs:")
def complex_function(required_arg, *args, default_arg="默认值", **kwargs):
    """复杂参数组合示例"""
    print(f"   必需参数: {required_arg}")
    print(f"   可变位置参数: {args}")
    print(f"   默认参数: {default_arg}")
    print(f"   可变关键字参数: {kwargs}")

# 测试复杂参数组合
print("   调用complex_function():")
complex_function("必需", 1, 2, 3, default_arg="自定义", extra1="额外1", extra2="额外2")

# 参数解包
print(f"\n📝 4. 参数解包:")

# 列表/元组解包为位置参数
def multiply(a, b, c):
    """三个数相乘"""
    return a * b * c

numbers = [2, 3, 4]
print(f"   列表解包: multiply(*{numbers}) = {multiply(*numbers)}")

numbers_tuple = (5, 6, 7)
print(f"   元组解包: multiply(*{numbers_tuple}) = {multiply(*numbers_tuple)}")

# 字典解包为关键字参数
def create_profile(name, age, city, hobby):
    """创建个人档案"""
    return f"{name}，{age}岁，来自{city}，爱好{hobby}"

profile_data = {
    "name": "李四",
    "age": 30,
    "city": "上海",
    "hobby": "编程"
}
print(f"   字典解包: create_profile(**{profile_data})")
print(f"   结果: {create_profile(**profile_data)}")

# 实际应用：灵活的配置函数
print(f"\n📝 5. 实际应用 - 灵活配置函数:")
def configure_system(*features, **settings):
    """系统配置函数"""
    print("   系统配置:")
    
    # 处理功能列表
    if features:
        print(f"     启用功能: {', '.join(features)}")
    
    # 处理设置字典
    if settings:
        print("     系统设置:")
        for key, value in settings.items():
            print(f"       {key}: {value}")
    
    return len(features), len(settings)

# 测试配置函数
feature_list = ["登录", "注册", "搜索"]
settings_dict = {
    "debug": True,
    "max_users": 1000,
    "timeout": 30
}

feature_count, setting_count = configure_system(*feature_list, **settings_dict)
print(f"   配置统计: {feature_count}个功能, {setting_count}个设置")

# 数据处理管道
print(f"\n📝 6. 数据处理管道:")
def data_pipeline(data, *operations, **options):
    """数据处理管道"""
    print(f"   原始数据: {data}")
    result = data
    
    # 应用操作
    for operation in operations:
        if operation == "sort":
            result = sorted(result)
        elif operation == "reverse":
            result = list(reversed(result))
        elif operation == "unique":
            result = list(set(result))
        elif operation == "filter_positive":
            result = [x for x in result if x > 0]
    
    # 应用选项
    if options.get("verbose", False):
        print(f"   处理步骤: {operations}")
        print(f"   处理选项: {options}")
    
    print(f"   处理结果: {result}")
    return result

# 测试数据管道
test_data = [3, -1, 4, 2, -5, 4, 3, 1]
processed = data_pipeline(
    test_data,
    "filter_positive", "unique", "sort",
    verbose=True,
    max_length=10
)

print(f"\n✅ 可变参数完成")
print(f"🎯 学习目标达成:")
print(f"   ✓ 掌握*args和**kwargs使用")
print(f"   ✓ 理解可变参数应用场景")
print(f"   ✓ 能设计灵活可变参数函数")
````

### 参数类型提示 [⭐基础]
**知识点说明**：类型提示提高了代码的可读性和可维护性，帮助IDE提供更好的代码补全和错误检查。

**学习要求**：

````python
print("📝 参数类型提示:")
print("=" * 50)

# 基本类型提示
print(f"📝 1. 基本类型提示:")
def add_numbers(a: int, b: int) -> int:
    """两个整数相加"""
    return a + b

def greet_user(name: str) -> str:
    """生成问候语"""
    return f"Hello, {name}!"

def calculate_discount(price: float, discount_rate: float) -> float:
    """计算折扣价格"""
    return price * (1 - discount_rate)

# 测试基本类型提示
print(f"   整数相加: {add_numbers(10, 20)}")
print(f"   问候语: {greet_user('张三')}")
print(f"   折扣价格: {calculate_discount(100.0, 0.2):.2f}")

# 复杂类型提示
print(f"\n📝 2. 复杂类型提示:")
from typing import List, Dict, Tuple, Optional, Union

def process_numbers(numbers: List[int]) -> List[int]:
    """处理整数列表"""
    return [x * 2 for x in numbers]

def create_user_info(name: str, age: int, scores: List[float]) -> Dict[str, Union[str, int, List[float]]]:
    """创建用户信息字典"""
    return {
        "name": name,
        "age": age,
        "scores": scores,
        "average": sum(scores) / len(scores) if scores else 0
    }

def find_coordinates(data: List[Tuple[int, int]]) -> Optional[Tuple[int, int]]:
    """查找特定坐标"""
    for x, y in data:
        if x == y:
            return (x, y)
    return None

# 测试复杂类型提示
numbers = [1, 2, 3, 4, 5]
processed = process_numbers(numbers)
print(f"   处理数字: {numbers} -> {processed}")

user_info = create_user_info("李四", 25, [85.5, 92.0, 78.5])
print(f"   用户信息: {user_info}")

coordinates = [(1, 2), (3, 3), (4, 5)]
found = find_coordinates(coordinates)
print(f"   查找坐标: {found}")

# 可选参数和默认值的类型提示
print(f"\n📝 3. 可选参数和默认值:")
def format_message(message: str, prefix: Optional[str] = None, suffix: str = "!") -> str:
    """格式化消息"""
    if prefix:
        message = f"{prefix}: {message}"
    return f"{message}{suffix}"

def calculate_stats(values: List[float], round_result: bool = True) -> Dict[str, float]:
    """计算统计信息"""
    if not values:
        return {"mean": 0.0, "std": 0.0}
    
    mean = sum(values) / len(values)
    variance = sum((x - mean) ** 2 for x in values) / len(values)
    std = variance ** 0.5
    
    if round_result:
        mean = round(mean, 2)
        std = round(std, 2)
    
    return {"mean": mean, "std": std}

# 测试可选参数
print(f"   格式化消息1: {format_message('Hello')}")
print(f"   格式化消息2: {format_message('Warning', 'ERROR')}")
print(f"   格式化消息3: {format_message('Info', suffix='.')}")

stats = calculate_stats([1.5, 2.5, 3.5, 4.5])
print(f"   统计信息: {stats}")

# 函数作为参数的类型提示
print(f"\n📝 4. 函数作为参数:")
from typing import Callable

def apply_operation(numbers: List[int], operation: Callable[[int], int]) -> List[int]:
    """对列表中的每个数字应用操作"""
    return [operation(num) for num in numbers]

def filter_list(items: List[str], condition: Callable[[str], bool]) -> List[str]:
    """根据条件过滤列表"""
    return [item for item in items if condition(item)]

# 定义操作函数
def square(x: int) -> int:
    return x * x

def is_long_word(word: str) -> bool:
    return len(word) > 4

# 测试函数参数
numbers = [1, 2, 3, 4, 5]
squared = apply_operation(numbers, square)
print(f"   平方操作: {numbers} -> {squared}")

words = ["hello", "world", "python", "is", "awesome"]
long_words = filter_list(words, is_long_word)
print(f"   长单词过滤: {long_words}")

# 类型别名
print(f"\n📝 5. 类型别名:")
from typing import TypeAlias

# 定义类型别名
UserID: TypeAlias = int
UserName: TypeAlias = str
UserRecord: TypeAlias = Dict[str, Union[UserID, UserName, str, int]]
ProcessorFunc: TypeAlias = Callable[[List[int]], List[int]]

def create_user_record(user_id: UserID, name: UserName, email: str, age: int) -> UserRecord:
    """创建用户记录"""
    return {
        "id": user_id,
        "name": name,
        "email": email,
        "age": age
    }

def batch_process(data: List[List[int]], processor: ProcessorFunc) -> List[List[int]]:
    """批量处理数据"""
    return [processor(batch) for batch in data]

# 测试类型别名
user = create_user_record(1001, "王五", "wangwu@example.com", 28)
print(f"   用户记录: {user}")

batches = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
processed_batches = batch_process(batches, lambda x: [i * 2 for i in x])
print(f"   批量处理: {processed_batches}")

# 泛型类型提示
print(f"\n📝 6. 泛型类型提示:")
from typing import TypeVar, Generic

# 定义类型变量
T = TypeVar('T')

def get_first_element(items: List[T]) -> Optional[T]:
    """获取列表第一个元素"""
    return items[0] if items else None

def find_max(items: List[T]) -> Optional[T]:
    """查找最大值"""
    return max(items) if items else None

# 测试泛型
first_num = get_first_element([10, 20, 30])
first_str = get_first_element(["apple", "banana", "cherry"])
print(f"   第一个数字: {first_num}")
print(f"   第一个字符串: {first_str}")

max_num = find_max([1, 5, 3, 9, 2])
max_str = find_max(["apple", "zebra", "banana"])
print(f"   最大数字: {max_num}")
print(f"   最大字符串: {max_str}")

print(f"\n✅ 参数类型提示完成")
print(f"🎯 学习目标达成:")
print(f"   ✓ 掌握基本类型提示语法")
print(f"   ✓ 理解复杂类型提示方法")
print(f"   ✓ 能编写规范类型提示代码")
````

### 参数传递的实际应用 [⭐基础]
**知识点说明**：通过实际案例展示参数传递机制在复杂场景中的应用，包括配置管理、数据处理和API设计。

````python
print("💼 参数传递的实际应用:")
print("=" * 50)

# 案例1：灵活的数据库查询接口
print(f"📝 1. 数据库查询接口:")
from typing import Any, Dict, List, Optional, Union

def build_query(table: str, 
                columns: Optional[List[str]] = None,
                where_conditions: Optional[Dict[str, Any]] = None,
                order_by: Optional[str] = None,
                limit: Optional[int] = None,
                **extra_params) -> str:
    """构建SQL查询语句"""
    
    # 基础SELECT部分
    if columns:
        columns_str = ', '.join(columns)
    else:
        columns_str = '*'
    
    query = f"SELECT {columns_str} FROM {table}"
    
    # WHERE条件
    if where_conditions:
        conditions = []
        for column, value in where_conditions.items():
            if isinstance(value, str):
                conditions.append(f"{column} = '{value}'")
            else:
                conditions.append(f"{column} = {value}")
        query += f" WHERE {' AND '.join(conditions)}"
    
    # ORDER BY
    if order_by:
        query += f" ORDER BY {order_by}"
    
    # LIMIT
    if limit:
        query += f" LIMIT {limit}"
    
    # 额外参数
    if extra_params:
        print(f"   额外参数: {extra_params}")
    
    return query

# 测试查询构建
print("   基础查询:")
query1 = build_query("users")
print(f"     {query1}")

print("\n   复杂查询:")
query2 = build_query(
    table="users",
    columns=["id", "name", "email"],
    where_conditions={"age": 25, "status": "active"},
    order_by="name",
    limit=10,
    debug=True
)
print(f"     {query2}")

# 案例2：灵活的API响应格式化
print(f"\n📝 2. API响应格式化:")

def format_api_response(data: Any,
                        status_code: int = 200,
                        message: str = "Success",
                        include_metadata: bool = False,
                        field_mapping: Optional[Dict[str, str]] = None,
                        **metadata) -> Dict[str, Any]:
    """格式化API响应"""
    
    response = {
        "status": status_code,
        "message": message,
        "data": data
    }
    
    # 字段映射
    if field_mapping and isinstance(data, dict):
        mapped_data = {}
        for old_key, new_key in field_mapping.items():
            if old_key in data:
                mapped_data[new_key] = data[old_key]
        response["data"] = mapped_data
    
    # 包含元数据
    if include_metadata:
        response["metadata"] = {
            "timestamp": "2025-11-30T22:45:00Z",
            "version": "1.0",
            **metadata
        }
    
    return response

# 测试API响应格式化
user_data = {"user_id": 123, "user_name": "张三", "user_email": "zhangsan@example.com"}

print("   基础响应:")
response1 = format_api_response(user_data)
print(f"     {response1}")

print("\n   完整响应:")
response2 = format_api_response(
    user_data,
    status_code=201,
    message="User created",
    include_metadata=True,
    field_mapping={"user_id": "id", "user_name": "name", "user_email": "email"},
    request_id="req_123",
    processing_time="0.05s"
)
print(f"     {response2}")

# 案例3：通用数据处理管道
print(f"\n📝 3. 通用数据处理管道:")

def create_data_pipeline(initial_data: List[Any],
                          *processors: Callable[[List[Any]], List[Any]],
                          validate_input: Optional[Callable[[List[Any]], bool]] = None,
                          error_handler: Optional[Callable[[Exception, List[Any]], List[Any]]] = None,
                          **config) -> List[Any]:
    """创建数据处理管道"""
    
    print(f"   开始处理数据，初始数量: {len(initial_data)}")
    
    # 输入验证
    if validate_input:
        if not validate_input(initial_data):
            raise ValueError("输入数据验证失败")
        print("   输入验证通过")
    
    # 配置信息
    if config:
        print(f"   配置参数: {config}")
    
    # 处理数据
    current_data = initial_data.copy()
    
    for i, processor in enumerate(processors, 1):
        try:
            print(f"   执行处理器 {i}: {processor.__name__ if hasattr(processor, '__name__') else '匿名函数'}")
            current_data = processor(current_data)
            print(f"     处理后数量: {len(current_data)}")
        except Exception as e:
            print(f"     处理器 {i} 出错: {e}")
            if error_handler:
                current_data = error_handler(e, current_data)
                print(f"     错误处理后数量: {len(current_data)}")
            else:
                raise
    
    print(f"   处理完成，最终数量: {len(current_data)}")
    return current_data

# 定义处理器函数
def remove_duplicates(data: List[Any]) -> List[Any]:
    return list(set(data))

def filter_positive(data: List[int]) -> List[int]:
    return [x for x in data if isinstance(x, int) and x > 0]

def sort_data(data: List[Any]) -> List[Any]:
    return sorted(data)

def validate_numbers(data: List[Any]) -> bool:
    return all(isinstance(x, (int, float)) for x in data)

def handle_error(error: Exception, data: List[Any]) -> List[Any]:
    print(f"     错误处理: {error}")
    return data  # 返回原始数据

# 测试数据管道
test_data = [1, 2, 2, -3, 4, 5, 5, 0, 6]
print("\n   数据管道测试:")

processed_data = create_data_pipeline(
    test_data,
    filter_positive,
    remove_duplicates,
    sort_data,
    validate_input=validate_numbers,
    error_handler=handle_error,
    verbose=True,
    max_retries=3
)

print(f"   最终结果: {processed_data}")

# 案例4：插件系统架构
print(f"\n📝 4. 插件系统架构:")

class PluginManager:
    def __init__(self):
        self.plugins = {}
    
    def register_plugin(self, name: str, plugin: Callable, **metadata):
        """注册插件"""
        self.plugins[name] = {
            "func": plugin,
            "metadata": metadata
        }
        print(f"   插件 '{name}' 已注册")
    
    def execute_plugin(self, name: str, *args, **kwargs):
        """执行插件"""
        if name not in self.plugins:
            raise ValueError(f"插件 '{name}' 未找到")
        
        plugin_info = self.plugins[name]
        print(f"   执行插件 '{name}': {plugin_info['metadata']}")
        
        return plugin_info["func"](*args, **kwargs)
    
    def execute_all_plugins(self, *args, **kwargs):
        """执行所有插件"""
        results = {}
        for name in self.plugins:
            try:
                results[name] = self.execute_plugin(name, *args, **kwargs)
            except Exception as e:
                results[name] = f"Error: {e}"
        return results

# 定义插件函数
def text_analyzer(text: str, min_length: int = 1) -> Dict[str, Any]:
    """文本分析插件"""
    words = [word for word in text.split() if len(word) >= min_length]
    return {
        "word_count": len(words),
        "char_count": len(text),
        "words": words[:5]  # 返回前5个词
    }

def number_calculator(*numbers: int) -> Dict[str, Union[int, float]]:
    """数字计算插件"""
    if not numbers:
        return {"error": "No numbers provided"}
    
    return {
        "count": len(numbers),
        "sum": sum(numbers),
        "average": sum(numbers) / len(numbers),
        "max": max(numbers),
        "min": min(numbers)
    }

def list_processor(items: List[Any], operation: str = "count") -> Dict[str, Any]:
    """列表处理插件"""
    if operation == "count":
        return {"count": len(items), "items": items}
    elif operation == "unique":
        return {"unique_count": len(set(items)), "unique_items": list(set(items))}
    else:
        return {"error": f"Unknown operation: {operation}"}

# 测试插件系统
print("\n   插件系统测试:")
plugin_manager = PluginManager()

# 注册插件
plugin_manager.register_plugin(
    "text_analyzer", 
    text_analyzer, 
    version="1.0", 
    author="AI Assistant",
    description="Analyze text content"
)

plugin_manager.register_plugin(
    "number_calculator", 
    number_calculator, 
    version="1.1", 
    author="AI Assistant",
    description="Calculate number statistics"
)

plugin_manager.register_plugin(
    "list_processor", 
    list_processor, 
    version="1.2", 
    author="AI Assistant",
    description="Process list data"
)

# 执行单个插件
print("\n   执行单个插件:")
text_result = plugin_manager.execute_plugin("text_analyzer", "Hello Python World!", min_length=3)
print(f"     文本分析结果: {text_result}")

number_result = plugin_manager.execute_plugin("number_calculator", 10, 20, 30, 40, 50)
print(f"     数字计算结果: {number_result}")

# 执行所有插件
print("\n   批量执行插件:")
all_results = plugin_manager.execute_all_plugins([1, 2, 3, 2, 1])
for plugin_name, result in all_results.items():
    print(f"     {plugin_name}: {result}")

print(f"\n✅ 参数传递的实际应用完成")
print(f"🎯 学习目标达成:")
print(f"   ✓ 掌握参数传递实际应用技巧")
print(f"   ✓ 理解灵活参数设计重要性")
print(f"   ✓ 能设计实用参数接口")
````

### 🔧 常见错误与注意事项

1. **参数顺序错误**：
   ```python
   # 错误：关键字参数在位置参数之前
   def func(a, b, c):
       pass
   func(b=2, 1, c=3)  # 语法错误
   
   # 正确：位置参数在前，关键字参数在后
   func(1, b=2, c=3)
   ```

2. **可变参数滥用**：
   ```python
   # 避免过度使用可变参数
   def bad_func(*args, **kwargs):
       # 参数含义不明确
       pass
   
   # 推荐使用明确的参数
   def good_func(data, options=None, verbose=False):
       # 参数含义清晰
       pass
   ```

3. **默认参数的陷阱**：
   ```python
   # 错误：使用可变对象作为默认参数
   def bad_func(items=[]):
       items.append(1)
       return items
   
   # 正确：使用None作为默认值
   def good_func(items=None):
       if items is None:
           items = []
       items.append(1)
       return items
   ```

4. **类型提示过度复杂**：
   ```python
   # 避免过度复杂的类型提示
   def overly_complex(data: Dict[str, Union[List[Dict[str, int]], Tuple[int, ...]]]) -> bool:
       pass
   
   # 使用类型别名简化
   from typing import TypeAlias
   
   ComplexType: TypeAlias = Dict[str, Union[List[Dict[str, int]], Tuple[int, ...]]]
   def simplified(data: ComplexType) -> bool:
       pass
   ```
