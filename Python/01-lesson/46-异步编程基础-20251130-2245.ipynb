````python
print("âš¡ å¼‚æ­¥ç¼–ç¨‹åŸºç¡€:")
print("=" * 50)

# å¯¼å…¥å¿…è¦çš„åº“
import asyncio
import time
import datetime
import random
import threading
from typing import List, Any, Optional, Callable
from dataclasses import dataclass
import json
import os
import requests
import aiohttp
import nest_asyncio

# å…è®¸åœ¨Jupyterä¸­è¿è¡Œå¼‚æ­¥ä»£ç 
nest_asyncio.apply()

print(f"âœ… Pythonç‰ˆæœ¬: {__import__('sys').version}")
print(f"âœ… asyncioç‰ˆæœ¬: {asyncio.__version__}")
print(f"âœ… å½“å‰æ—¶é—´: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# 1. å¼‚æ­¥ç¼–ç¨‹åŸºç¡€æ¦‚å¿µ
print(f"\nğŸ“ 1. å¼‚æ­¥ç¼–ç¨‹åŸºç¡€æ¦‚å¿µ:")

# 1.1 åŒæ­¥ä¸å¼‚æ­¥çš„å¯¹æ¯”
print(f"\n   ğŸ”„ 1.1 åŒæ­¥ä¸å¼‚æ­¥çš„å¯¹æ¯”:")

def sync_task(name: str, duration: float) -> str:
    """åŒæ­¥ä»»åŠ¡"""
    print(f"      å¼€å§‹åŒæ­¥ä»»åŠ¡ {name}...")
    time.sleep(duration)
    print(f"      å®ŒæˆåŒæ­¥ä»»åŠ¡ {name}")
    return f"åŒæ­¥ä»»åŠ¡ {name} å®Œæˆ"

async def async_task(name: str, duration: float) -> str:
    """å¼‚æ­¥ä»»åŠ¡"""
    print(f"      å¼€å§‹å¼‚æ­¥ä»»åŠ¡ {name}...")
    await asyncio.sleep(duration)
    print(f"      å®Œæˆå¼‚æ­¥ä»»åŠ¡ {name}")
    return f"å¼‚æ­¥ä»»åŠ¡ {name} å®Œæˆ"

# åŒæ­¥æ‰§è¡Œç¤ºä¾‹
print(f"\n   ğŸ”¸ åŒæ­¥æ‰§è¡Œç¤ºä¾‹:")
start_time = time.time()

result1 = sync_task("ä»»åŠ¡A", 1)
result2 = sync_task("ä»»åŠ¡B", 1)
result3 = sync_task("ä»»åŠ¡C", 1)

sync_time = time.time() - start_time
print(f"   åŒæ­¥æ‰§è¡Œæ€»æ—¶é—´: {sync_time:.2f} ç§’")
print(f"   ç»“æœ: {result1}, {result2}, {result3}")

# å¼‚æ­¥æ‰§è¡Œç¤ºä¾‹
print(f"\n   ğŸ”¸ å¼‚æ­¥æ‰§è¡Œç¤ºä¾‹:")

async def run_async_example():
    start_time = time.time()
    
    # å¹¶å‘æ‰§è¡Œå¼‚æ­¥ä»»åŠ¡
    task1 = async_task("ä»»åŠ¡A", 1)
    task2 = async_task("ä»»åŠ¡B", 1)
    task3 = async_task("ä»»åŠ¡C", 1)
    
    # ç­‰å¾…æ‰€æœ‰ä»»åŠ¡å®Œæˆ
    results = await asyncio.gather(task1, task2, task3)
    
    async_time = time.time() - start_time
    print(f"   å¼‚æ­¥æ‰§è¡Œæ€»æ—¶é—´: {async_time:.2f} ç§’")
    print(f"   ç»“æœ: {', '.join(results)}")
    print(f"   æ€§èƒ½æå‡: {(sync_time - async_time) / sync_time * 100:.1f}%")
    
    return results, async_time

# è¿è¡Œå¼‚æ­¥ç¤ºä¾‹
async_results, async_time = await run_async_example()

# 1.2 åç¨‹åŸºç¡€
print(f"\n   ğŸŒ€ 1.2 åç¨‹åŸºç¡€:")

async def simple_coroutine():
    """ç®€å•åç¨‹ç¤ºä¾‹"""
    print("      åç¨‹å¼€å§‹æ‰§è¡Œ")
    print("      åç¨‹æ‰§è¡Œä¸­...")
    print("      åç¨‹æ‰§è¡Œå®Œæˆ")
    return "åç¨‹ç»“æœ"

async def coroutine_with_params(name: str, age: int):
    """å¸¦å‚æ•°çš„åç¨‹"""
    print(f"      ä½ å¥½ï¼Œæˆ‘æ˜¯ {name}ï¼Œä»Šå¹´ {age} å²")
    await asyncio.sleep(0.1)
    return f"{name} çš„ä¿¡æ¯å·²å¤„ç†"

# è¿è¡Œç®€å•åç¨‹
print(f"\n   ğŸ”¸ è¿è¡Œç®€å•åç¨‹:")
result = await simple_coroutine()
print(f"   åç¨‹è¿”å›ç»“æœ: {result}")

# è¿è¡Œå¸¦å‚æ•°çš„åç¨‹
print(f"\n   ğŸ”¸ è¿è¡Œå¸¦å‚æ•°çš„åç¨‹:")
result = await coroutine_with_params("å¼ ä¸‰", 25)
print(f"   åç¨‹è¿”å›ç»“æœ: {result}")

# 1.3 async/await è¯­æ³•è¯¦è§£
print(f"\n   ğŸ”§ 1.3 async/await è¯­æ³•è¯¦è§£:")

async def demonstrate_async_await():
    """æ¼”ç¤ºasync/awaitè¯­æ³•"""
    print(f"      1. è¿›å…¥å¼‚æ­¥å‡½æ•°")
    
    # await åªèƒ½åœ¨asyncå‡½æ•°ä¸­ä½¿ç”¨
    print(f"      2. ç­‰å¾…å¼‚æ­¥æ“ä½œ...")
    await asyncio.sleep(0.5)
    print(f"      3. å¼‚æ­¥æ“ä½œå®Œæˆ")
    
    # å¯ä»¥awaitå…¶ä»–åç¨‹
    print(f"      4. è°ƒç”¨å…¶ä»–åç¨‹...")
    sub_result = await async_task("å­ä»»åŠ¡", 0.3)
    print(f"      5. å­ä»»åŠ¡ç»“æœ: {sub_result}")
    
    return "async/awaitæ¼”ç¤ºå®Œæˆ"

print(f"\n   ğŸ”¸ async/awaitè¯­æ³•æ¼”ç¤º:")
result = await demonstrate_async_await()
print(f"   æœ€ç»ˆç»“æœ: {result}")

print(f"\nâœ… å¼‚æ­¥ç¼–ç¨‹åŸºç¡€æ¦‚å¿µå®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å¼‚æ­¥ç¼–ç¨‹çš„åŸºæœ¬æ¦‚å¿µå’ŒåŸç†")
print(f"   âœ“ ç†è§£åŒæ­¥ä¸å¼‚æ­¥ç¼–ç¨‹çš„åŒºåˆ«")
print(f"   âœ“ ç†Ÿç»ƒä½¿ç”¨async/awaitè¯­æ³•")
print(f"   âœ“ èƒ½å¤Ÿç¼–å†™åŸºç¡€çš„å¼‚æ­¥ç¨‹åº")
````

### å¼‚æ­¥ä»»åŠ¡ç®¡ç† [â­åŸºç¡€]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šå¼‚æ­¥ä»»åŠ¡ç®¡ç†æ˜¯å¼‚æ­¥ç¼–ç¨‹çš„æ ¸å¿ƒæŠ€èƒ½ï¼ŒåŒ…æ‹¬ä»»åŠ¡åˆ›å»ºã€è°ƒåº¦ã€ç­‰å¾…å’Œå–æ¶ˆã€‚æŒæ¡å¼‚æ­¥ä»»åŠ¡ç®¡ç†å¯¹äºæ„å»ºå¤æ‚çš„LangChainåº”ç”¨éå¸¸é‡è¦ã€‚

**å­¦ä¹ è¦æ±‚**ï¼š
- æŒæ¡å¼‚æ­¥ä»»åŠ¡çš„åˆ›å»ºå’Œç®¡ç†
- ç†è§£ä»»åŠ¡è°ƒåº¦å’Œæ‰§è¡Œæœºåˆ¶
- ç†Ÿç»ƒä½¿ç”¨asyncio.gatherç­‰å·¥å…·
- èƒ½å¤Ÿå¤„ç†å¼‚æ­¥ä»»åŠ¡çš„å¼‚å¸¸

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°å®Œæ•´çš„å¼‚æ­¥ä»»åŠ¡ç®¡ç†ç³»ç»Ÿ
- è¿›è¡Œä»»åŠ¡è°ƒåº¦çš„æ€§èƒ½ä¼˜åŒ–
- åº”ç”¨ä»»åŠ¡ç®¡ç†è§£å†³å®é™…é—®é¢˜
- éªŒè¯ç‚¹ï¼šèƒ½ç‹¬ç«‹æ„å»ºæœ‰æ•ˆçš„ä»»åŠ¡ç®¡ç†ç³»ç»Ÿ

````python
print("\nğŸ“‹ å¼‚æ­¥ä»»åŠ¡ç®¡ç†:")
print("=" * 50)

# 1. å¼‚æ­¥ä»»åŠ¡åˆ›å»ºå’Œç®¡ç†
print(f"ğŸ“ 1. å¼‚æ­¥ä»»åŠ¡åˆ›å»ºå’Œç®¡ç†:")

async def create_and_manage_tasks():
    """åˆ›å»ºå’Œç®¡ç†å¼‚æ­¥ä»»åŠ¡"""
    print(f"   ğŸ”¸ åˆ›å»ºå¼‚æ­¥ä»»åŠ¡:")
    
    # åˆ›å»ºä»»åŠ¡ä½†ä¸ç«‹å³æ‰§è¡Œ
    task1 = asyncio.create_task(async_task("ä»»åŠ¡1", 0.5))
    task2 = asyncio.create_task(async_task("ä»»åŠ¡2", 0.8))
    task3 = asyncio.create_task(async_task("ä»»åŠ¡3", 0.3))
    
    print(f"      å·²åˆ›å»º 3 ä¸ªä»»åŠ¡")
    print(f"      ä»»åŠ¡1çŠ¶æ€: {task1.done()}")
    print(f"      ä»»åŠ¡2çŠ¶æ€: {task2.done()}")
    print(f"      ä»»åŠ¡3çŠ¶æ€: {task3.done()}")
    
    # ç­‰å¾…å•ä¸ªä»»åŠ¡
    print(f"\n   ğŸ”¸ ç­‰å¾…ä»»åŠ¡å®Œæˆ:")
    result1 = await task1
    print(f"      ä»»åŠ¡1å®Œæˆ: {result1}")
    print(f"      ä»»åŠ¡1çŠ¶æ€: {task1.done()}")
    
    # ç­‰å¾…å‰©ä½™ä»»åŠ¡
    result2, result3 = await asyncio.gather(task2, task3)
    print(f"      ä»»åŠ¡2å®Œæˆ: {result2}")
    print(f"      ä»»åŠ¡3å®Œæˆ: {result3}")
    
    return [result1, result2, result3]

# è¿è¡Œä»»åŠ¡ç®¡ç†ç¤ºä¾‹
task_results = await create_and_manage_tasks()

# 2. ä»»åŠ¡è°ƒåº¦å’Œæ‰§è¡Œ
print(f"\n   ğŸ”„ 2. ä»»åŠ¡è°ƒåº¦å’Œæ‰§è¡Œ:")

async def batch_task_processor(task_names: List[str], delay: float = 0.1):
    """æ‰¹é‡ä»»åŠ¡å¤„ç†å™¨"""
    print(f"   å¤„ç† {len(task_names)} ä¸ªä»»åŠ¡...")
    
    # åˆ›å»ºæ‰€æœ‰ä»»åŠ¡
    tasks = [asyncio.create_task(async_task(name, delay)) for name in task_names]
    
    # ç­‰å¾…æ‰€æœ‰ä»»åŠ¡å®Œæˆ
    results = await asyncio.gather(*tasks)
    
    print(f"   æ‰¹é‡å¤„ç†å®Œæˆ: {len(results)} ä¸ªç»“æœ")
    return results

# æ‰¹é‡å¤„ç†ä»»åŠ¡
batch_results = await batch_task_processor(["æ‰¹é‡ä»»åŠ¡1", "æ‰¹é‡ä»»åŠ¡2", "æ‰¹é‡ä»»åŠ¡3", "æ‰¹é‡ä»»åŠ¡4"], 0.2)

# 3. ä»»åŠ¡è¶…æ—¶å’Œå–æ¶ˆ
print(f"\n   â±ï¸ 3. ä»»åŠ¡è¶…æ—¶å’Œå–æ¶ˆ:")

async def long_running_task(name: str, duration: float):
    """é•¿æ—¶é—´è¿è¡Œçš„ä»»åŠ¡"""
    try:
        print(f"      å¼€å§‹é•¿æ—¶é—´ä»»åŠ¡ {name} (é¢„è®¡ {duration}s)...")
        await asyncio.sleep(duration)
        print(f"      é•¿æ—¶é—´ä»»åŠ¡ {name} å®Œæˆ")
        return f"{name} å®Œæˆ"
    except asyncio.CancelledError:
        print(f"      é•¿æ—¶é—´ä»»åŠ¡ {name} è¢«å–æ¶ˆ")
        raise

async def demonstrate_timeout_and_cancel():
    """æ¼”ç¤ºè¶…æ—¶å’Œå–æ¶ˆ"""
    print(f"   ğŸ”¸ ä»»åŠ¡è¶…æ—¶æ¼”ç¤º:")
    
    try:
        # è®¾ç½®2ç§’è¶…æ—¶
        result = await asyncio.wait_for(long_running_task("è¶…æ—¶ä»»åŠ¡", 3), timeout=2)
        print(f"      ä»»åŠ¡å®Œæˆ: {result}")
    except asyncio.TimeoutError:
        print(f"      ä»»åŠ¡è¶…æ—¶ï¼")
    
    print(f"\n   ğŸ”¸ ä»»åŠ¡å–æ¶ˆæ¼”ç¤º:")
    
    # åˆ›å»ºä»»åŠ¡
    task = asyncio.create_task(long_running_task("å¯å–æ¶ˆä»»åŠ¡", 3))
    
    # ç­‰å¾…1ç§’åå–æ¶ˆ
    await asyncio.sleep(1)
    
    if not task.done():
        print(f"      å–æ¶ˆä»»åŠ¡...")
        task.cancel()
        
        try:
            await task
        except asyncio.CancelledError:
            print(f"      ä»»åŠ¡å·²å–æ¶ˆ")

# æ¼”ç¤ºè¶…æ—¶å’Œå–æ¶ˆ
await demonstrate_timeout_and_cancel()

# 4. å¼‚æ­¥ä»»åŠ¡å¼‚å¸¸å¤„ç†
print(f"\n   âš ï¸ 4. å¼‚æ­¥ä»»åŠ¡å¼‚å¸¸å¤„ç†:")

async def task_with_exception(name: str, should_fail: bool = False):
    """å¯èƒ½æŠ›å‡ºå¼‚å¸¸çš„ä»»åŠ¡"""
    print(f"      æ‰§è¡Œä»»åŠ¡ {name}...")
    await asyncio.sleep(0.1)
    
    if should_fail:
        raise ValueError(f"ä»»åŠ¡ {name} æ‰§è¡Œå¤±è´¥")
    
    return f"ä»»åŠ¡ {name} æˆåŠŸå®Œæˆ"

async def demonstrate_exception_handling():
    """æ¼”ç¤ºå¼‚å¸¸å¤„ç†"""
    print(f"   ğŸ”¸ å•ä¸ªä»»åŠ¡å¼‚å¸¸å¤„ç†:")
    
    try:
        result = await task_with_exception("å¤±è´¥ä»»åŠ¡", should_fail=True)
        print(f"      ç»“æœ: {result}")
    except ValueError as e:
        print(f"      æ•è·å¼‚å¸¸: {e}")
    
    print(f"\n   ğŸ”¸ å¤šä»»åŠ¡å¼‚å¸¸å¤„ç†:")
    
    # åˆ›å»ºå¤šä¸ªä»»åŠ¡ï¼Œå…¶ä¸­ä¸€äº›ä¼šå¤±è´¥
    tasks = [
        asyncio.create_task(task_with_exception("æˆåŠŸä»»åŠ¡1", False)),
        asyncio.create_task(task_with_exception("å¤±è´¥ä»»åŠ¡2", True)),
        asyncio.create_task(task_with_exception("æˆåŠŸä»»åŠ¡3", False)),
    ]
    
    # ä½¿ç”¨gatherå¤„ç†å¼‚å¸¸
    try:
        results = await asyncio.gather(*tasks, return_exceptions=True)
        
        print(f"      å¤„ç† {len(results)} ä¸ªç»“æœ:")
        for i, result in enumerate(results):
            if isinstance(result, Exception):
                print(f"        ä»»åŠ¡{i+1}: å¼‚å¸¸ - {result}")
            else:
                print(f"        ä»»åŠ¡{i+1}: æˆåŠŸ - {result}")
    
    except Exception as e:
        print(f"      æœªé¢„æœŸçš„å¼‚å¸¸: {e}")

# æ¼”ç¤ºå¼‚å¸¸å¤„ç†
await demonstrate_exception_handling()

print(f"\nâœ… å¼‚æ­¥ä»»åŠ¡ç®¡ç†å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å¼‚æ­¥ä»»åŠ¡çš„åˆ›å»ºå’Œç®¡ç†")
print(f"   âœ“ ç†è§£ä»»åŠ¡è°ƒåº¦å’Œæ‰§è¡Œæœºåˆ¶")
print(f"   âœ“ ç†Ÿç»ƒä½¿ç”¨asyncio.gatherç­‰å·¥å…·")
print(f"   âœ“ èƒ½å¤Ÿå¤„ç†å¼‚æ­¥ä»»åŠ¡çš„å¼‚å¸¸")
````
