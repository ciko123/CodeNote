### é›†åˆæ•°å­¦è¿ç®— [â­åŸºç¡€]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šé›†åˆæ”¯æŒå„ç§æ•°å­¦è¿ç®—ï¼Œå¦‚å¹¶é›†ã€äº¤é›†ã€å·®é›†å’Œå¯¹ç§°å·®é›†ã€‚

**å­¦ä¹ è¦æ±‚**ï¼š
- æŒæ¡é›†åˆçš„åŸºæœ¬æ•°å­¦è¿ç®—
- ç†è§£é›†åˆè¿ç®—çš„å«ä¹‰å’Œåº”ç”¨
- èƒ½çµæ´»è¿ç”¨é›†åˆè¿ç®—è§£å†³é—®é¢˜

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°é›†åˆæ•°å­¦è¿ç®—
- è¿›è¡Œå¤æ‚é›†åˆæ“ä½œ
- åº”ç”¨é›†åˆè¿ç®—è§£å†³å®é™…é—®é¢˜
- éªŒè¯ç‚¹ï¼šèƒ½æ­£ç¡®ä½¿ç”¨é›†åˆæ•°å­¦è¿ç®—

````python
print("ğŸ§® é›†åˆæ•°å­¦è¿ç®—:")
print("=" * 50)

# å‡†å¤‡æµ‹è¯•é›†åˆ
print(f"ğŸ“ 1. å‡†å¤‡æµ‹è¯•é›†åˆ:")
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
set_c = {1, 2, 3}

print(f"   é›†åˆA: {set_a}")
print(f"   é›†åˆB: {set_b}")
print(f"   é›†åˆC: {set_c}")

# å¹¶é›†è¿ç®—
print(f"\nğŸ“ 2. å¹¶é›†è¿ç®— (Union |):")
union_ab = set_a | set_b
union_method = set_a.union(set_b)
union_multiple = set_a | set_b | set_c

print(f"   A | B: {union_ab}")
print(f"   A.union(B): {union_method}")
print(f"   A | B | C: {union_multiple}")

# äº¤é›†è¿ç®—
print(f"\nğŸ“ 3. äº¤é›†è¿ç®— (Intersection &):")
intersection_ab = set_a & set_b
intersection_method = set_a.intersection(set_b)
intersection_multiple = set_a & set_b & set_c

print(f"   A & B: {intersection_ab}")
print(f"   A.intersection(B): {intersection_method}")
print(f"   A & B & C: {intersection_multiple}")

# å·®é›†è¿ç®—
print(f"\nğŸ“ 4. å·®é›†è¿ç®— (Difference -):")
diff_ab = set_a - set_b
diff_method = set_a.difference(set_b)
diff_ba = set_b - set_a

print(f"   A - B: {diff_ab}")
print(f"   A.difference(B): {diff_method}")
print(f"   B - A: {diff_ba}")

# å¯¹ç§°å·®é›†è¿ç®—
print(f"\nğŸ“ 5. å¯¹ç§°å·®é›†è¿ç®— (Symmetric Difference ^):")
sym_diff_ab = set_a ^ set_b
sym_diff_method = set_a.symmetric_difference(set_b)

print(f"   A ^ B: {sym_diff_ab}")
print(f"   A.symmetric_difference(B): {sym_diff_method}")

# å­é›†å’Œè¶…é›†åˆ¤æ–­
print(f"\nğŸ“ 6. å­é›†å’Œè¶…é›†åˆ¤æ–­:")
print(f"   Cæ˜¯Açš„å­é›†: {set_c.issubset(set_a)}")
print(f"   Aæ˜¯Cçš„è¶…é›†: {set_a.issuperset(set_c)}")
print(f"   Aå’ŒBä¸ç›¸äº¤: {set_a.isdisjoint(set_b)}")

# é›†åˆè¿ç®—çš„å®é™…åº”ç”¨
print(f"\nğŸ“ 7. é›†åˆè¿ç®—çš„å®é™…åº”ç”¨:")

# å­¦ç”Ÿé€‰è¯¾ç³»ç»Ÿ
math_students = {"å¼ ä¸‰", "æå››", "ç‹äº”", "èµµå…­", "é’±ä¸ƒ"}
physics_students = {"æå››", "ç‹äº”", "å­™å…«", "å‘¨ä¹", "å´å"}
chemistry_students = {"å¼ ä¸‰", "å­™å…«", "éƒ‘åä¸€", "ç‹äº”", "å†¯åäºŒ"}

print(f"   æ•°å­¦è¯¾å­¦ç”Ÿ: {math_students}")
print(f"   ç‰©ç†è¯¾å­¦ç”Ÿ: {physics_students}")
print(f"   åŒ–å­¦è¯¾å­¦ç”Ÿ: {chemistry_students}")

# é€‰ä¿®å¤šé—¨è¯¾ç¨‹çš„å­¦ç”Ÿ
math_physics = math_students & physics_students
all_three = math_students & physics_students & chemistry_students
any_course = math_students | physics_students | chemistry_students

print(f"\n   åŒæ—¶é€‰ä¿®æ•°å­¦å’Œç‰©ç†: {math_physics}")
print(f"   åŒæ—¶é€‰ä¿®ä¸‰é—¨è¯¾ç¨‹: {all_three}")
print(f"   é€‰ä¿®è‡³å°‘ä¸€é—¨è¯¾ç¨‹: {any_course}")

# åªé€‰ä¿®ä¸€é—¨è¯¾ç¨‹çš„å­¦ç”Ÿ
only_math = math_students - physics_students - chemistry_students
only_physics = physics_students - math_students - chemistry_students
only_chemistry = chemistry_students - math_students - physics_students
only_one = only_math | only_physics | only_chemistry

print(f"\n   åªé€‰ä¿®æ•°å­¦: {only_math}")
print(f"   åªé€‰ä¿®ç‰©ç†: {only_physics}")
print(f"   åªé€‰ä¿®åŒ–å­¦: {only_chemistry}")
print(f"   åªé€‰ä¿®ä¸€é—¨è¯¾ç¨‹: {only_one}")

# æ•°æ®å»é‡å’Œåˆ†æ
print(f"\nğŸ“ 8. æ•°æ®å»é‡å’Œåˆ†æ:")

# æ¨¡æ‹Ÿç½‘ç«™è®¿é—®æ—¥å¿—
page_visits_day1 = ["é¦–é¡µ", "äº§å“é¡µ", "é¦–é¡µ", "å…³äºé¡µ", "äº§å“é¡µ", "é¦–é¡µ"]
page_visits_day2 = ["é¦–é¡µ", "äº§å“é¡µ", "è”ç³»é¡µ", "äº§å“é¡µ", "åšå®¢é¡µ", "é¦–é¡µ"]
page_visits_day3 = ["å…³äºé¡µ", "äº§å“é¡µ", "é¦–é¡µ", "åšå®¢é¡µ", "é¦–é¡µ", "è”ç³»é¡µ"]

# è½¬æ¢ä¸ºé›†åˆå»é‡
unique_pages_day1 = set(page_visits_day1)
unique_pages_day2 = set(page_visits_day2)
unique_pages_day3 = set(page_visits_day3)

print(f"   ç¬¬1å¤©è®¿é—®é¡µé¢: {unique_pages_day1}")
print(f"   ç¬¬2å¤©è®¿é—®é¡µé¢: {unique_pages_day2}")
print(f"   ç¬¬3å¤©è®¿é—®é¡µé¢: {unique_pages_day3}")

# åˆ†æè®¿é—®æ¨¡å¼
all_pages = unique_pages_day1 | unique_pages_day2 | unique_pages_day3
common_pages = unique_pages_day1 & unique_pages_day2 & unique_pages_day3
unique_to_day1 = unique_pages_day1 - unique_pages_day2 - unique_pages_day3

print(f"\n   ä¸‰å¤©æ€»å…±è®¿é—®é¡µé¢: {all_pages}")
print(f"   ä¸‰å¤©éƒ½è®¿é—®çš„é¡µé¢: {common_pages}")
print(f"   åªæœ‰ç¬¬1å¤©è®¿é—®çš„é¡µé¢: {unique_to_day1}")

# æƒé™ç®¡ç†ç³»ç»Ÿ
print(f"\nğŸ“ 9. æƒé™ç®¡ç†ç³»ç»Ÿ:")

admin_permissions = {"read", "write", "delete", "manage_users", "system_config"}
editor_permissions = {"read", "write", "edit_content", "publish"}
viewer_permissions = {"read"}

print(f"   ç®¡ç†å‘˜æƒé™: {admin_permissions}")
print(f"   ç¼–è¾‘æƒé™: {editor_permissions}")
print(f"   æŸ¥çœ‹è€…æƒé™: {viewer_permissions}")

# æƒé™æ£€æŸ¥
def check_permission(user_permissions, required_permission):
    return required_permission in user_permissions

# æƒé™å‡çº§
def upgrade_permissions(base_permissions, additional_permissions):
    return base_permissions | additional_permissions

# æƒé™é™çº§
def downgrade_permissions(base_permissions, remove_permissions):
    return base_permissions - remove_permissions

print(f"\n   æƒé™æ£€æŸ¥ç¤ºä¾‹:")
print(f"     ç¼–è¾‘è€…èƒ½å¦åˆ é™¤: {check_permission(editor_permissions, 'delete')}")
print(f"     ç®¡ç†å‘˜èƒ½å¦å†™å…¥: {check_permission(admin_permissions, 'write')}")

print(f"\n   æƒé™å‡çº§ç¤ºä¾‹:")
upgraded_editor = upgrade_permissions(editor_permissions, {"delete"})
print(f"     ç¼–è¾‘è€…è·å¾—åˆ é™¤æƒé™å: {upgraded_editor}")

print(f"\n   æƒé™é™çº§ç¤ºä¾‹:")
limited_admin = downgrade_permissions(admin_permissions, {"system_config"})
print(f"     ç®¡ç†å‘˜å¤±å»ç³»ç»Ÿé…ç½®æƒé™å: {limited_admin}")

print(f"\nâœ… é›†åˆæ•°å­¦è¿ç®—å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡é›†åˆåŸºæœ¬æ•°å­¦è¿ç®—")
print(f"   âœ“ ç†è§£é›†åˆè¿ç®—å«ä¹‰å’Œåº”ç”¨")
print(f"   âœ“ èƒ½çµæ´»è¿ç”¨é›†åˆè¿ç®—è§£å†³é—®é¢˜")
````

- æŒæ¡å­—å…¸å’Œé›†åˆçš„ç»„åˆä½¿ç”¨
- ç†å¤æ‚æ•°æ®ç»“æ„çš„å¤„ç†æ€è·¯
- èƒ½ç‹¬ç«‹è®¾è®¡å­—å…¸é›†åˆè§£å†³æ–¹æ¡ˆ

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°å®Œæ•´çš„å­—å…¸é›†åˆåº”ç”¨
- å¤„ç†å¤æ‚çš„æ•°æ®ç®¡ç†ä»»åŠ¡
- ä¼˜åŒ–æ•°æ®æ“ä½œæ•ˆç‡

````python
print("ğŸ¯ å­—å…¸é›†åˆç»¼åˆåº”ç”¨:")
print("=" * 50)

# æ¡ˆä¾‹1ï¼šç”¨æˆ·ç®¡ç†ç³»ç»Ÿ
print(f"ğŸ“ 1. ç”¨æˆ·ç®¡ç†ç³»ç»Ÿ:")

class UserManager:
    def __init__(self):
        self.users = {}  # ç”¨æˆ·ID -> ç”¨æˆ·ä¿¡æ¯
        self.permissions = {}  # ç”¨æˆ·ID -> æƒé™é›†åˆ
        self.groups = {}  # ç»„å -> ç”¨æˆ·é›†åˆ
    
    def add_user(self, user_id, name, email, role):
        """æ·»åŠ ç”¨æˆ·"""
        self.users[user_id] = {
            'name': name,
            'email': email,
            'role': role,
            'created_at': '2024-01-01'
        }
        self.permissions[user_id] = set()
        return self.users[user_id]
    
    def grant_permission(self, user_id, permission):
        """æˆäºˆæƒé™"""
        if user_id in self.permissions:
            self.permissions[user_id].add(permission)
    
    def revoke_permission(self, user_id, permission):
        """æ’¤é”€æƒé™"""
        if user_id in self.permissions:
            self.permissions[user_id].discard(permission)
    
    def add_to_group(self, user_id, group_name):
        """æ·»åŠ åˆ°ç»„"""
        if group_name not in self.groups:
            self.groups[group_name] = set()
        self.groups[group_name].add(user_id)
    
    def get_users_with_permission(self, permission):
        """è·å–å…·æœ‰ç‰¹å®šæƒé™çš„ç”¨æˆ·"""
        return {user_id for user_id, perms in self.permissions.items() 
                if permission in perms}
    
    def get_group_users(self, group_name):
        """è·å–ç»„å†…ç”¨æˆ·"""
        return self.groups.get(group_name, set())
    
    def get_user_summary(self, user_id):
        """è·å–ç”¨æˆ·æ‘˜è¦"""
        if user_id not in self.users:
            return None
        
        user_info = self.users[user_id].copy()
        user_info['permissions'] = list(self.permissions.get(user_id, set()))
        user_info['groups'] = [group for group, users in self.groups.items() 
                             if user_id in users]
        return user_info

# æµ‹è¯•ç”¨æˆ·ç®¡ç†ç³»ç»Ÿ
manager = UserManager()

# æ·»åŠ ç”¨æˆ·
users_data = [
    (1, "å¼ ä¸‰", "zhang@example.com", "admin"),
    (2, "æå››", "li@example.com", "editor"),
    (3, "ç‹äº”", "wang@example.com", "viewer"),
    (4, "èµµå…­", "zhao@example.com", "editor")
]

for user_id, name, email, role in users_data:
    user = manager.add_user(user_id, name, email, role)
    print(f"   æ·»åŠ ç”¨æˆ·: {name} (ID: {user_id}, è§’è‰²: {role})")

# æˆäºˆæƒé™
manager.grant_permission(1, {"read", "write", "delete", "manage_users"})
manager.grant_permission(2, {"read", "write", "edit_content"})
manager.grant_permission(3, {"read"})
manager.grant_permission(4, {"read", "write", "publish"})

# æ·»åŠ åˆ°ç»„
manager.add_to_group(1, "ç®¡ç†å‘˜")
manager.add_to_group(2, "ç¼–è¾‘ç»„")
manager.add_to_group(4, "ç¼–è¾‘ç»„")
manager.add_to_group(3, "æŸ¥çœ‹ç»„")

# æŸ¥è¯¢ç”¨æˆ·
print(f"\n   ç”¨æˆ·è¯¦ç»†ä¿¡æ¯:")
for user_id in [1, 2, 3, 4]:
    summary = manager.get_user_summary(user_id)
    print(f"     {summary['name']}: æƒé™{summary['permissions']}, ç»„{summary['groups']}")

# æƒé™æŸ¥è¯¢
write_users = manager.get_users_with_permission("write")
print(f"\n   å…·æœ‰å†™å…¥æƒé™çš„ç”¨æˆ·: {write_users}")

editors = manager.get_group_users("ç¼–è¾‘ç»„")
print(f"   ç¼–è¾‘ç»„æˆå‘˜: {editors}")

# æ¡ˆä¾‹2ï¼šå•†å“æ¨èç³»ç»Ÿ
print(f"\nğŸ“ 2. å•†å“æ¨èç³»ç»Ÿ:")

class ProductRecommendation:
    def __init__(self):
        self.user_preferences = {}  # ç”¨æˆ·ID -> å…´è¶£é›†åˆ
        self.product_categories = {}  # å•†å“ID -> åˆ†ç±»é›†åˆ
        self.user_purchases = {}  # ç”¨æˆ·ID -> è´­ä¹°å•†å“é›†åˆ
        self.category_products = {}  # åˆ†ç±» -> å•†å“é›†åˆ
    
    def add_user_preference(self, user_id, interests):
        """æ·»åŠ ç”¨æˆ·å…´è¶£"""
        self.user_preferences[user_id] = set(interests)
    
    def add_product(self, product_id, name, categories):
        """æ·»åŠ å•†å“"""
        self.product_categories[product_id] = {
            'name': name,
            'categories': set(categories)
        }
        
        # æ›´æ–°åˆ†ç±»ç´¢å¼•
        for category in categories:
            if category not in self.category_products:
                self.category_products[category] = set()
            self.category_products[category].add(product_id)
    
    def record_purchase(self, user_id, product_id):
        """è®°å½•è´­ä¹°"""
        if user_id not in self.user_purchases:
            self.user_purchases[user_id] = set()
        self.user_purchases[user_id].add(product_id)
    
    def recommend_products(self, user_id, limit=5):
        """æ¨èå•†å“"""
        if user_id not in self.user_preferences:
            return []
        
        user_interests = self.user_preferences[user_id]
        purchased = self.user_purchases.get(user_id, set())
        
        # æ‰¾åˆ°ç›¸å…³å•†å“
        candidate_products = set()
        for interest in user_interests:
            if interest in self.category_products:
                candidate_products.update(self.category_products[interest])
        
        # æ’é™¤å·²è´­ä¹°å•†å“
        recommendations = candidate_products - purchased
        
        # è½¬æ¢ä¸ºå•†å“ä¿¡æ¯åˆ—è¡¨
        recommended_items = []
        for product_id in recommendations:
            if product_id in self.product_categories:
                product_info = self.product_categories[product_id]
                recommended_items.append({
                    'id': product_id,
                    'name': product_info['name'],
                    'categories': list(product_info['categories'])
                })
        
        return recommended_items[:limit]
    
    def get_similar_users(self, user_id, threshold=0.3):
        """è·å–ç›¸ä¼¼ç”¨æˆ·"""
        if user_id not in self.user_preferences:
            return set()
        
        user_interests = self.user_preferences[user_id]
        similar_users = set()
        
        for other_id, other_interests in self.user_preferences.items():
            if other_id != user_id:
                # è®¡ç®—å…´è¶£ç›¸ä¼¼åº¦ï¼ˆJaccardç›¸ä¼¼ç³»æ•°ï¼‰
                intersection = len(user_interests & other_interests)
                union = len(user_interests | other_interests)
                similarity = intersection / union if union > 0 else 0
                
                if similarity >= threshold:
                    similar_users.add(other_id)
        
        return similar_users

# æµ‹è¯•æ¨èç³»ç»Ÿ
recommender = ProductRecommendation()

# æ·»åŠ å•†å“
products_data = [
    (101, "ç¬”è®°æœ¬ç”µè„‘", ["ç”µå­äº§å“", "åŠå…¬"])
    (102, "æ— çº¿é¼ æ ‡", ["ç”µå­äº§å“", "åŠå…¬", "æ¸¸æˆ"]),
    (103, "æœºæ¢°é”®ç›˜", ["ç”µå­äº§å“", "åŠå…¬", "æ¸¸æˆ"]),
    (104, "è¿åŠ¨é‹", ["æœè£…", "è¿åŠ¨"]),
    (105, "è·‘æ­¥æœº", ["è¿åŠ¨", "å¥èº«"]),
    (106, "ç‘œä¼½å«", ["è¿åŠ¨", "å¥èº«"]),
    (107, "ç¼–ç¨‹ä¹¦ç±", ["å›¾ä¹¦", "æ•™è‚²"]),
    (108, "å°è¯´", ["å›¾ä¹¦", "å¨±ä¹"])
]

for product_id, name, categories in products_data:
    recommender.add_product(product_id, name, categories)
    print(f"   æ·»åŠ å•†å“: {name} (ID: {product_id}, åˆ†ç±»: {categories})")

# æ·»åŠ ç”¨æˆ·å…´è¶£
users_interests = {
    1: ["ç”µå­äº§å“", "åŠå…¬", "æ¸¸æˆ"],
    2: ["è¿åŠ¨", "å¥èº«"],
    3: ["å›¾ä¹¦", "æ•™è‚²"],
    4: ["ç”µå­äº§å“", "åŠå…¬", "å›¾ä¹¦"]
}

for user_id, interests in users_interests.items():
    recommender.add_user_preference(user_id, interests)
    print(f"\n   ç”¨æˆ·{user_id}å…´è¶£: {interests}")

# è®°å½•è´­ä¹°å†å²
recommender.record_purchase(1, 101)  # ç”¨æˆ·1ä¹°äº†ç¬”è®°æœ¬ç”µè„‘
recommender.record_purchase(2, 104)  # ç”¨æˆ·2ä¹°äº†è¿åŠ¨é‹
recommender.record_purchase(3, 107)  # ç”¨æˆ·3ä¹°äº†ç¼–ç¨‹ä¹¦ç±

# ç”Ÿæˆæ¨è
print(f"\n   å•†å“æ¨è:")
for user_id in [1, 2, 3, 4]:
    recommendations = recommender.recommend_products(user_id, 3)
    print(f"\n     ç”¨æˆ·{user_id}æ¨èå•†å“:")
    for item in recommendations:
        print(f"       - {item['name']} (åˆ†ç±»: {', '.join(item['categories'])})")

# æ‰¾ç›¸ä¼¼ç”¨æˆ·
print(f"\n   ç›¸ä¼¼ç”¨æˆ·åˆ†æ:")
similar_to_1 = recommender.get_similar_users(1, 0.2)
similar_to_4 = recommender.get_similar_users(4, 0.2)
print(f"   ä¸ç”¨æˆ·1ç›¸ä¼¼çš„ç”¨æˆ·: {similar_to_1}")
print(f"   ä¸ç”¨æˆ·4ç›¸ä¼¼çš„ç”¨æˆ·: {similar_to_4}")

# æ¡ˆä¾‹3ï¼šæ•°æ®åˆ†æç®¡é“
print(f"\nğŸ“ 3. æ•°æ®åˆ†æç®¡é“:")

def data_analysis_pipeline(raw_data):
    """æ•°æ®åˆ†æç®¡é“"""
    print(f"   åŸå§‹æ•°æ®è®°å½•æ•°: {len(raw_data)}")
    
    # æ­¥éª¤1: æ•°æ®æ¸…æ´—å’Œå»é‡
    unique_records = []
    seen_ids = set()
    
    for record in raw_data:
        record_id = record.get('id')
        if record_id and record_id not in seen_ids:
            seen_ids.add(record_id)
            unique_records.append(record)
    
    print(f"   æ­¥éª¤1-å»é‡åè®°å½•æ•°: {len(unique_records)}")
    
    # æ­¥éª¤2: æ•°æ®åˆ†ç±»ç»Ÿè®¡
    category_stats = {}
    category_sets = {}
    
    for record in unique_records:
        category = record.get('category', 'æœªçŸ¥')
        if category not in category_stats:
            category_stats[category] = {'count': 0, 'total_value': 0}
            category_sets[category] = set()
        
        category_stats[category]['count'] += 1
        category_stats[category]['total_value'] += record.get('value', 0)
        category_sets[category].add(record['id'])
    
    print(f"\n   æ­¥éª¤2-åˆ†ç±»ç»Ÿè®¡:")
    for category, stats in category_stats.items():
        avg_value = stats['total_value'] / stats['count']
        print(f"     {category}: {stats['count']}æ¡è®°å½•, æ€»å€¼{stats['total_value']}, å¹³å‡{avg_value:.2f}")
    
    # æ­¥éª¤3: å¼‚å¸¸æ£€æµ‹
    all_values = [record.get('value', 0) for record in unique_records]
    if all_values:
        avg_value = sum(all_values) / len(all_values)
        std_dev = (sum((x - avg_value) ** 2 for x in all_values) / len(all_values)) ** 0.5
        
        outliers = []
        for record in unique_records:
            value = record.get('value', 0)
            if abs(value - avg_value) > 2 * std_dev:  # 2å€æ ‡å‡†å·®
                outliers.append(record)
        
        print(f"\n   æ­¥éª¤3-å¼‚å¸¸æ£€æµ‹:")
        print(f"     å¹³å‡å€¼: {avg_value:.2f}, æ ‡å‡†å·®: {std_dev:.2f}")
        print(f"     å¼‚å¸¸è®°å½•æ•°: {len(outliers)}")
        for outlier in outliers:
            print(f"       ID{outlier['id']}: å€¼{outlier['value']} (ç±»åˆ«: {outlier['category']})")
    
    # æ­¥éª¤4: æ•°æ®è´¨é‡æŠ¥å‘Š
    quality_report = {
        'total_records': len(unique_records),
        'categories': len(category_stats),
        'duplicate_ids': len(raw_data) - len(unique_records),
        'completeness': {},
        'unique_values': {}
    }
    
    # è®¡ç®—å®Œæ•´æ€§
    for field in ['name', 'category', 'value']:
        complete_count = sum(1 for record in unique_records if field in record and record[field] is not None)
        quality_report['completeness'][field] = complete_count / len(unique_records) if unique_records else 0
    
    # è®¡ç®—å”¯ä¸€å€¼æ•°é‡
    for field in ['name', 'category']:
        unique_values = set(record.get(field) for record in unique_records if record.get(field))
        quality_report['unique_values'][field] = len(unique_values)
    
    print(f"\n   æ­¥éª¤4-æ•°æ®è´¨é‡æŠ¥å‘Š:")
    print(f"     æ€»è®°å½•æ•°: {quality_report['total_records']}")
    print(f"     åˆ†ç±»æ•°: {quality_report['categories']}")
    print(f"     é‡å¤è®°å½•: {quality_report['duplicate_ids']}")
    print(f"     å­—æ®µå®Œæ•´æ€§: {quality_report['completeness']}")
    print(f"     å”¯ä¸€å€¼æ•°é‡: {quality_report['unique_values']}")
    
    return {
        'clean_data': unique_records,
        'category_stats': category_stats,
        'quality_report': quality_report
    }

# æµ‹è¯•æ•°æ®åˆ†æç®¡é“
test_data = [
    {'id': 1, 'name': 'äº§å“A', 'category': 'ç”µå­äº§å“', 'value': 100},
    {'id': 2, 'name': 'äº§å“B', 'category': 'æœè£…', 'value': 50},
    {'id': 3, 'name': 'äº§å“C', 'category': 'ç”µå­äº§å“', 'value': 150},
    {'id': 1, 'name': 'äº§å“A', 'category': 'ç”µå­äº§å“', 'value': 100},  # é‡å¤
    {'id': 4, 'name': 'äº§å“D', 'category': 'é£Ÿå“', 'value': 30},
    {'id': 5, 'name': 'äº§å“E', 'category': 'æœè£…', 'value': 80},
    {'id': 6, 'name': 'äº§å“F', 'category': 'ç”µå­äº§å“', 'value': 500},  # å¼‚å¸¸å€¼
    {'id': 7, 'name': None, 'category': 'æœè£…', 'value': 60},  # ç¼ºå¤±åç§°
    {'id': 8, 'category': 'é£Ÿå“', 'value': 25},  # ç¼ºå¤±åç§°
    {'id': 9, 'name': 'äº§å“I', 'category': None, 'value': 40}  # ç¼ºå¤±åˆ†ç±»
]

analysis_result = data_analysis_pipeline(test_data)

print(f"\nâœ… å­—å…¸é›†åˆç»¼åˆåº”ç”¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å­—å…¸å’Œé›†åˆç»„åˆä½¿ç”¨")
print(f"   âœ“ ç†è§£å¤æ‚æ•°æ®ç»“æ„å¤„ç†")
print(f"   âœ“ èƒ½ç‹¬ç«‹è®¾è®¡å­—å…¸é›†åˆè§£å†³æ–¹æ¡ˆ")
````
