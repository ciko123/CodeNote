```python
# æ¡ˆä¾‹1ï¼šå­¦ç”Ÿæˆç»©ç®¡ç†

class StudentManager:
    def __init__(self):
        self.students = []
    
    def add_student(self, name, scores):
        """æ·»åŠ å­¦ç”Ÿ"""
        student = {
            'name': name,
            'scores': scores,
            'average': sum(scores) / len(scores),
            'grade': self._calculate_grade(sum(scores) / len(scores))
        }
        self.students.append(student)
        return student
    
    def _calculate_grade(self, average):
        """è®¡ç®—ç­‰çº§"""
        if average >= 90:
            return 'A'
        elif average >= 80:
            return 'B'
        elif average >= 70:
            return 'C'
        elif average >= 60:
            return 'D'
        else:
            return 'F'
    
    def get_top_students(self, n=3):
        """è·å–å‰Nåå­¦ç”Ÿ"""
        sorted_students = sorted(self.students, key=lambda x: x['average'], reverse=True)
        return sorted_students[:n]
    
    def get_grade_statistics(self):
        """è·å–ç­‰çº§ç»Ÿè®¡"""
        grades = [student['grade'] for student in self.students]
        grade_counts = {}
        for grade in ['A', 'B', 'C', 'D', 'F']:
            grade_counts[grade] = grades.count(grade)
        return grade_counts

# æµ‹è¯•å­¦ç”Ÿç®¡ç†ç³»ç»Ÿ
manager = StudentManager()

# æ·»åŠ å­¦ç”Ÿ
students_data = [
    ("å¼ ä¸‰", [85, 92, 78, 88, 90]),
    ("æå››", [92, 88, 95, 86, 91]),
    ("ç‹äº”", [78, 85, 82, 80, 79]),
    ("èµµå…­", [95, 93, 89, 94, 92]),
    ("é’±ä¸ƒ", [72, 68, 75, 70, 73])
]

for name, scores in students_data:
    student = manager.add_student(name, scores)
    print(f"   æ·»åŠ å­¦ç”Ÿ: {name}, å¹³å‡åˆ†: {student['average']:.1f}, ç­‰çº§: {student['grade']}")

# è·å–å‰ä¸‰å
top_students = manager.get_top_students(3)
print(f"\n   å‰ä¸‰åå­¦ç”Ÿ:")
for i, student in enumerate(top_students, 1):
    print(f"     ç¬¬{i}å: {student['name']} ({student['average']:.1f}åˆ†, {student['grade']}çº§)")

# ç­‰çº§ç»Ÿè®¡
grade_stats = manager.get_grade_statistics()
print(f"\n   ç­‰çº§ç»Ÿè®¡: {grade_stats}")

# æ¡ˆä¾‹2ï¼šåº“å­˜ç®¡ç†ç³»ç»Ÿ
print(f"\nğŸ“ 2. åº“å­˜ç®¡ç†ç³»ç»Ÿ:")

class InventoryManager:
    def __init__(self):
        self.products = []
    
    def add_product(self, name, price, quantity, category):
        """æ·»åŠ å•†å“"""
        product = {
            'id': len(self.products) + 1,
            'name': name,
            'price': price,
            'quantity': quantity,
            'category': category,
            'total_value': price * quantity
        }
        self.products.append(product)
        return product
    
    def update_quantity(self, product_id, new_quantity):
        """æ›´æ–°åº“å­˜æ•°é‡"""
        for product in self.products:
            if product['id'] == product_id:
                product['quantity'] = new_quantity
                product['total_value'] = product['price'] * new_quantity
                return True
        return False
    
    def get_low_stock_products(self, threshold=10):
        """è·å–ä½åº“å­˜å•†å“"""
        return [product for product in self.products if product['quantity'] < threshold]
    
    def get_category_summary(self):
        """è·å–åˆ†ç±»æ±‡æ€»"""
        categories = {}
        for product in self.products:
            category = product['category']
            if category not in categories:
                categories[category] = {'count': 0, 'total_value': 0}
            categories[category]['count'] += 1
            categories[category]['total_value'] += product['total_value']
        return categories

# æµ‹è¯•åº“å­˜ç®¡ç†
inventory = InventoryManager()

# æ·»åŠ å•†å“
products_data = [
    ("ç¬”è®°æœ¬ç”µè„‘", 5999, 15, "ç”µå­äº§å“"),
    ("æ— çº¿é¼ æ ‡", 99, 8, "ç”µå­äº§å“"),
    ("åŠå…¬æ¤…", 299, 5, "å®¶å…·"),
    ("ä¹¦æ¡Œ", 899, 12, "å®¶å…·"),
    ("å’–å•¡æœº", 1299, 3, "å®¶ç”µ")
]

for name, price, quantity, category in products_data:
    product = inventory.add_product(name, price, quantity, category)
    print(f"   æ·»åŠ å•†å“: {name} (ID: {product['id']}, ä»·æ ¼: Â¥{price}, åº“å­˜: {quantity})")

# æ›´æ–°åº“å­˜
inventory.update_quantity(2, 25)  # å¢åŠ é¼ æ ‡åº“å­˜
print(f"\n   æ›´æ–°æ— çº¿é¼ æ ‡åº“å­˜ä¸º25")

# ä½åº“å­˜è­¦å‘Š
low_stock = inventory.get_low_stock_products(10)
print(f"\n   ä½åº“å­˜å•†å“ (<10):")
for product in low_stock:
    print(f"     {product['name']}: {product['quantity']}ä¸ª")

# åˆ†ç±»æ±‡æ€»
category_summary = inventory.get_category_summary()
print(f"\n   åˆ†ç±»æ±‡æ€»:")
for category, summary in category_summary.items():
    print(f"     {category}: {summary['count']}ç§å•†å“, æ€»ä»·å€¼Â¥{summary['total_value']}")

# æ¡ˆä¾‹3ï¼šæ•°æ®åˆ†æå¤„ç†
print(f"\nğŸ“ 3. æ•°æ®åˆ†æå¤„ç†:")

def analyze_sales_data(sales_records):
    """åˆ†æé”€å”®æ•°æ®"""
    print(f"   é”€å”®è®°å½•æ•°é‡: {len(sales_records)}")
    
    # æå–æ‰€æœ‰é”€å”®é¢
    sales_amounts = [record['amount'] for record in sales_records]
    
    # åŸºç¡€ç»Ÿè®¡
    total_sales = sum(sales_amounts)
    average_sale = total_sales / len(sales_amounts)
    max_sale = max(sales_amounts)
    min_sale = min(sales_amounts)
    
    # æŒ‰äº§å“åˆ†ç±»
    product_sales = {}
    for record in sales_records:
        product = record['product']
        if product not in product_sales:
            product_sales[product] = []
        product_sales[product].append(record['amount'])
    
    # è®¡ç®—å„äº§å“ç»Ÿè®¡
    product_stats = {}
    for product, amounts in product_sales.items():
        product_stats[product] = {
            'count': len(amounts),
            'total': sum(amounts),
            'average': sum(amounts) / len(amounts),
            'max': max(amounts),
            'min': min(amounts)
        }
    
    # é”€å”®è¶‹åŠ¿åˆ†æ
    monthly_sales = {}
    for record in sales_records:
        month = record['date'][:7]  # æå–å¹´æœˆ
        if month not in monthly_sales:
            monthly_sales[month] = 0
        monthly_sales[month] += record['amount']
    
    return {
        'total_sales': total_sales,
        'average_sale': average_sale,
        'max_sale': max_sale,
        'min_sale': min_sale,
        'product_stats': product_stats,
        'monthly_sales': monthly_sales
    }

# æµ‹è¯•é”€å”®æ•°æ®åˆ†æ
sales_data = [
    {'date': '2024-01-15', 'product': 'ç¬”è®°æœ¬ç”µè„‘', 'amount': 5999},
    {'date': '2024-01-20', 'product': 'æ— çº¿é¼ æ ‡', 'amount': 99},
    {'date': '2024-01-25', 'product': 'ç¬”è®°æœ¬ç”µè„‘', 'amount': 5999},
    {'date': '2024-02-10', 'product': 'åŠå…¬æ¤…', 'amount': 299},
    {'date': '2024-02-15', 'product': 'æ— çº¿é¼ æ ‡', 'amount': 99},
    {'date': '2024-02-20', 'product': 'ä¹¦æ¡Œ', 'amount': 899},
    {'date': '2024-03-05', 'product': 'å’–å•¡æœº', 'amount': 1299},
    {'date': '2024-03-10', 'product': 'ç¬”è®°æœ¬ç”µè„‘', 'amount': 5999}
]

analysis_result = analyze_sales_data(sales_data)

print(f"\n   é”€å”®åˆ†æç»“æœ:")
print(f"     æ€»é”€å”®é¢: Â¥{analysis_result['total_sales']}")
print(f"     å¹³å‡é”€å”®é¢: Â¥{analysis_result['average_sale']:.2f}")
print(f"     æœ€é«˜é”€å”®é¢: Â¥{analysis_result['max_sale']}")
print(f"     æœ€ä½é”€å”®é¢: Â¥{analysis_result['min_sale']}")

print(f"\n   äº§å“é”€å”®ç»Ÿè®¡:")
for product, stats in analysis_result['product_stats'].items():
    print(f"     {product}: {stats['count']}æ¬¡é”€å”®, æ€»é¢Â¥{stats['total']}, å¹³å‡Â¥{stats['average']:.2f}")

print(f"\n   æœˆåº¦é”€å”®è¶‹åŠ¿:")
for month, total in analysis_result['monthly_sales'].items():
    print(f"     {month}: Â¥{total}")

# æ¡ˆä¾‹4ï¼šæ•°æ®æ¸…æ´—å’Œé¢„å¤„ç†
print(f"\nğŸ“ 4. æ•°æ®æ¸…æ´—å’Œé¢„å¤„ç†:")

def clean_data_pipeline(raw_data):
    """æ•°æ®æ¸…æ´—ç®¡é“"""
    print(f"   åŸå§‹æ•°æ®: {raw_data}")
    
    # æ­¥éª¤1: ç§»é™¤ç©ºå€¼å’ŒNone
    step1 = [item for item in raw_data if item is not None and item != ""]
    print(f"   æ­¥éª¤1-ç§»é™¤ç©ºå€¼: {step1}")
    
    # æ­¥éª¤2: æ ‡å‡†åŒ–å­—ç¬¦ä¸²ï¼ˆå»é™¤ç©ºæ ¼ï¼Œè½¬å°å†™ï¼‰
    step2 = []
    for item in step1:
        if isinstance(item, str):
            step2.append(item.strip().lower())
        else:
            step2.append(item)
    print(f"   æ­¥éª¤2-æ ‡å‡†åŒ–å­—ç¬¦ä¸²: {step2}")
    
    # æ­¥éª¤3: ç§»é™¤é‡å¤é¡¹
    step3 = []
    seen = set()
    for item in step2:
        if item not in seen:
            step3.append(item)
            seen.add(item)
    print(f"   æ­¥éª¤3-ç§»é™¤é‡å¤: {step3}")
    
    # æ­¥éª¤4: æ•°æ®ç±»å‹è½¬æ¢
    step4 = []
    for item in step3:
        if isinstance(item, str) and item.isdigit():
            step4.append(int(item))
        elif isinstance(item, str) and item.replace('.', '', 1).isdigit():
            step4.append(float(item))
        else:
            step4.append(item)
    print(f"   æ­¥éª¤4-ç±»å‹è½¬æ¢: {step4}")
    
    # æ­¥éª¤5: å¼‚å¸¸å€¼æ£€æµ‹
    step5 = []
    for item in step4:
        if isinstance(item, (int, float)):
            if abs(item) > 1000000:  # è¿‡å¤§å€¼
                step5.append(None)  # æ ‡è®°ä¸ºå¼‚å¸¸
            else:
                step5.append(item)
        else:
            step5.append(item)
    print(f"   æ­¥éª¤5-å¼‚å¸¸å€¼æ£€æµ‹: {step5}")
    
    return step5

# æµ‹è¯•æ•°æ®æ¸…æ´—
messy_data = [
    "Apple", " banana ", "", None, "APPLE", "123", "45.67", 
    "orange", "BANANA", "9999999", "  grape  ", "apple", "3.14"
]

cleaned_data = clean_data_pipeline(messy_data)
print(f"\n   æœ€ç»ˆæ¸…æ´—ç»“æœ: {cleaned_data}")
```
