````python
print("ğŸ”§ é«˜çº§æ—¥æœŸæ—¶é—´æ“ä½œ:")
print("=" * 50)

import datetime
from datetime import datetime, date, time, timedelta
import calendar
from dateutil import parser, relativedelta, rrule
import pytz
import holidays
from typing import List, Tuple, Optional

# 1. å·¥ä½œæ—¥è®¡ç®—
print(f"ğŸ“ 1. å·¥ä½œæ—¥è®¡ç®—:")

# 1.1 åŸºæœ¬å·¥ä½œæ—¥åˆ¤æ–­
print(f"\n   ğŸ’¼ 1.1 åŸºæœ¬å·¥ä½œæ—¥åˆ¤æ–­:")

def is_workday(dt: datetime) -> bool:
    """åˆ¤æ–­æ˜¯å¦ä¸ºå·¥ä½œæ—¥(å‘¨ä¸€åˆ°å‘¨äº”)"""
    return dt.weekday() < 5  # 0-4ä¸ºå‘¨ä¸€åˆ°å‘¨äº”

def is_weekend(dt: datetime) -> bool:
    """åˆ¤æ–­æ˜¯å¦ä¸ºå‘¨æœ«"""
    return dt.weekday() >= 5  # 5-6ä¸ºå‘¨å…­å‘¨æ—¥

def get_weekday_name(dt: datetime) -> str:
    """è·å–æ˜ŸæœŸåç§°"""
    weekdays = ['å‘¨ä¸€', 'å‘¨äºŒ', 'å‘¨ä¸‰', 'å‘¨å››', 'å‘¨äº”', 'å‘¨å…­', 'å‘¨æ—¥']
    return weekdays[dt.weekday()]

# æµ‹è¯•å·¥ä½œæ—¥åˆ¤æ–­
test_dates = [
    datetime(2024, 12, 23),  # å‘¨ä¸€
    datetime(2024, 12, 24),  # å‘¨äºŒ
    datetime(2024, 12, 25),  # å‘¨ä¸‰
    datetime(2024, 12, 28),  # å‘¨å…­
    datetime(2024, 12, 29),  # å‘¨æ—¥
]

for test_date in test_dates:
    weekday_name = get_weekday_name(test_date)
    is_work = is_workday(test_date)
    is_week = is_weekend(test_date)
    print(f"     {test_date.strftime('%Y-%m-%d')} {weekday_name}: {'å·¥ä½œæ—¥' if is_work else 'éå·¥ä½œæ—¥'} ({'å‘¨æœ«' if is_week else 'å·¥ä½œæ—¥'})")

# 1.2 å·¥ä½œæ—¥è®¡ç®—
print(f"\n   ğŸ“Š 1.2 å·¥ä½œæ—¥è®¡ç®—:")

def count_workdays(start_date: date, end_date: date) -> int:
    """è®¡ç®—ä¸¤ä¸ªæ—¥æœŸä¹‹é—´çš„å·¥ä½œæ—¥æ•°é‡"""
    workdays = 0
    current_date = start_date
    
    while current_date <= end_date:
        if is_workday(datetime.combine(current_date, time.min)):
            workdays += 1
        current_date += timedelta(days=1)
    
    return workdays

def count_weekends(start_date: date, end_date: date) -> int:
    """è®¡ç®—ä¸¤ä¸ªæ—¥æœŸä¹‹é—´çš„å‘¨æœ«å¤©æ•°"""
    weekends = 0
    current_date = start_date
    
    while current_date <= end_date:
        if is_weekend(datetime.combine(current_date, time.min)):
            weekends += 1
        current_date += timedelta(days=1)
    
    return weekends

def get_next_workday(dt: datetime) -> datetime:
    """è·å–ä¸‹ä¸€ä¸ªå·¥ä½œæ—¥"""
    next_day = dt + timedelta(days=1)
    while not is_workday(next_day):
        next_day += timedelta(days=1)
    return next_day

def get_previous_workday(dt: datetime) -> datetime:
    """è·å–ä¸Šä¸€ä¸ªå·¥ä½œæ—¥"""
    prev_day = dt - timedelta(days=1)
    while not is_workday(prev_day):
        prev_day -= timedelta(days=1)
    return prev_day

# æµ‹è¯•å·¥ä½œæ—¥è®¡ç®—
start = date(2024, 12, 1)
end = date(2024, 12, 31)

total_days = (end - start).days + 1
workdays = count_workdays(start, end)
weekends = count_weekends(start, end)

print(f"     2024å¹´12æœˆç»Ÿè®¡:")
print(f"     æ€»å¤©æ•°: {total_days}")
print(f"     å·¥ä½œæ—¥: {workdays}")
print(f"     å‘¨æœ«: {weekends}")

test_dt = datetime(2024, 12, 25)  # å‘¨ä¸‰
next_work = get_next_workday(test_dt)
prev_work = get_previous_workday(test_dt)

print(f"\n     æµ‹è¯•æ—¥æœŸ: {test_dt.strftime('%Y-%m-%d')} ({get_weekday_name(test_dt)})")
print(f"     ä¸‹ä¸€ä¸ªå·¥ä½œæ—¥: {next_work.strftime('%Y-%m-%d')} ({get_weekday_name(next_work)})")
print(f"     ä¸Šä¸€ä¸ªå·¥ä½œæ—¥: {prev_work.strftime('%Y-%m-%d')} ({get_weekday_name(prev_work)})")

# 1.3 èŠ‚å‡æ—¥å¤„ç†
print(f"\n   ğŸ‰ 1.3 èŠ‚å‡æ—¥å¤„ç†:")

def get_chinese_holidays(year: int) -> holidays.HolidayBase:
    """è·å–ä¸­å›½èŠ‚å‡æ—¥"""
    try:
        cn_holidays = holidays.China(years=year)
        return cn_holidays
    except Exception:
        # å¦‚æœholidaysåº“ä¸å¯ç”¨ï¼Œåˆ›å»ºç®€å•çš„èŠ‚å‡æ—¥åˆ—è¡¨
        simple_holidays = holidays.HolidayBase()
        # æ·»åŠ ä¸€äº›å›ºå®šèŠ‚å‡æ—¥
        simple_holidays[f'{year}-01-01'] = 'å…ƒæ—¦'
        simple_holidays[f'{year}-05-01'] = 'åŠ³åŠ¨èŠ‚'
        simple_holidays[f'{year}-10-01'] = 'å›½åº†èŠ‚'
        simple_holidays[f'{year}-10-02'] = 'å›½åº†èŠ‚'
        simple_holidays[f'{year}-10-03'] = 'å›½åº†èŠ‚'
        return simple_holidays

def is_holiday(dt: date, holidays_obj: holidays.HolidayBase) -> bool:
    """åˆ¤æ–­æ˜¯å¦ä¸ºèŠ‚å‡æ—¥"""
    return dt in holidays_obj

def is_business_day(dt: date, holidays_obj: holidays.HolidayBase) -> bool:
    """åˆ¤æ–­æ˜¯å¦ä¸ºè¥ä¸šæ—¥(å·¥ä½œæ—¥ä¸”éèŠ‚å‡æ—¥)"""
    return is_workday(datetime.combine(dt, time.min)) and not is_holiday(dt, holidays_obj)

# æµ‹è¯•èŠ‚å‡æ—¥
cn_holidays = get_chinese_holidays(2024)

test_dates = [
    date(2024, 1, 1),   # å…ƒæ—¦
    date(2024, 5, 1),   # åŠ³åŠ¨èŠ‚
    date(2024, 10, 1),  # å›½åº†èŠ‚
    date(2024, 12, 25), # æ™®é€šå·¥ä½œæ—¥
]

print(f"     èŠ‚å‡æ—¥æµ‹è¯•:")
for test_date in test_dates:
    is_hol = is_holiday(test_date, cn_holidays)
    is_bus = is_business_day(test_date, cn_holidays)
    holiday_name = cn_holidays.get(test_date, 'æ™®é€šæ—¥')
    print(f"     {test_date}: {holiday_name} ({'èŠ‚å‡æ—¥' if is_hol else 'éèŠ‚å‡æ—¥'}, {'è¥ä¸šæ—¥' if is_bus else 'éè¥ä¸šæ—¥'})")

# 2. å‘¨æœŸæ€§æ—¶é—´å¤„ç†
print(f"\nğŸ“ 2. å‘¨æœŸæ€§æ—¶é—´å¤„ç†:")

# 2.1 ä½¿ç”¨rruleç”Ÿæˆå‘¨æœŸæ€§æ—¥æœŸ
print(f"\n   ğŸ”„ 2.1 ä½¿ç”¨rruleç”Ÿæˆå‘¨æœŸæ€§æ—¥æœŸ:")

# æ¯å‘¨ç”Ÿæˆ
weekly_dates = list(rrule.rrule(
    rrule.WEEKLY,
    dtstart=datetime(2024, 1, 1),
    until=datetime(2024, 1, 31),
    byweekday=rrule.MO  # æ¯å‘¨ä¸€
))

print(f"     2024å¹´1æœˆçš„å‘¨ä¸€:")
for dt in weekly_dates[:5]:  # æ˜¾ç¤ºå‰5ä¸ª
    print(f"       {dt.strftime('%Y-%m-%d')} {get_weekday_name(dt)}")

# æ¯æœˆç”Ÿæˆ
monthly_dates = list(rrule.rrule(
    rrule.MONTHLY,
    dtstart=datetime(2024, 1, 1),
    until=datetime(2024, 12, 31),
    bymonthday=15  # æ¯æœˆ15æ—¥
))

print(f"\n     2024å¹´æ¯æœˆ15æ—¥:")
for dt in monthly_dates[:5]:
    print(f"       {dt.strftime('%Y-%m-%d')} {get_weekday_name(dt)}")

# æ¯å·¥ä½œæ—¥ç”Ÿæˆ
daily_business = list(rrule.rrule(
    rrule.DAILY,
    dtstart=datetime(2024, 12, 1),
    until=datetime(2024, 12, 10),
    byweekday=(rrule.MO, rrule.TU, rrule.WE, rrule.TH, rrule.FR)  # å‘¨ä¸€åˆ°å‘¨äº”
))

print(f"\n     2024å¹´12æœˆå‰10å¤©çš„å·¥ä½œæ—¥:")
for dt in daily_business:
    print(f"       {dt.strftime('%Y-%m-%d')} {get_weekday_name(dt)}")

# 2.2 è‡ªå®šä¹‰å‘¨æœŸç”Ÿæˆå™¨
print(f"\n   ğŸ”§ 2.2 è‡ªå®šä¹‰å‘¨æœŸç”Ÿæˆå™¨:")

class DateGenerator:
    """æ—¥æœŸç”Ÿæˆå™¨"""
    
    def __init__(self, start_date: date):
        self.start_date = start_date
    
    def generate_daily(self, count: int) -> List[date]:
        """ç”Ÿæˆè¿ç»­æ—¥æœŸ"""
        dates = []
        current_date = self.start_date
        
        for _ in range(count):
            dates.append(current_date)
            current_date += timedelta(days=1)
        
        return dates
    
    def generate_weekly(self, count: int, weekday: int = 0) -> List[date]:
        """ç”Ÿæˆæ¯å‘¨æŒ‡å®šæ˜ŸæœŸå‡ çš„æ—¥æœŸ"""
        dates = []
        current_date = self.start_date
        
        # æ‰¾åˆ°ç¬¬ä¸€ä¸ªæŒ‡å®šçš„æ˜ŸæœŸå‡ 
        while current_date.weekday() != weekday:
            current_date += timedelta(days=1)
        
        for _ in range(count):
            dates.append(current_date)
            current_date += timedelta(weeks=1)
        
        return dates
    
    def generate_monthly(self, count: int, day: int = 1) -> List[date]:
        """ç”Ÿæˆæ¯æœˆæŒ‡å®šæ—¥æœŸ"""
        dates = []
        current_date = self.start_date
        
        # è°ƒæ•´åˆ°æŒ‡å®šæ—¥æœŸ
        if current_date.day > day:
            current_date += relativedelta.relativedelta(months=1)
        
        current_date = current_date.replace(day=day)
        
        for _ in range(count):
            dates.append(current_date)
            current_date += relativedelta.relativedelta(months=1)
        
        return dates
    
    def generate_workdays(self, count: int) -> List[date]:
        """ç”Ÿæˆå·¥ä½œæ—¥"""
        dates = []
        current_date = self.start_date
        
        while len(dates) < count:
            if is_workday(datetime.combine(current_date, time.min)):
                dates.append(current_date)
            current_date += timedelta(days=1)
        
        return dates

# æµ‹è¯•æ—¥æœŸç”Ÿæˆå™¨
generator = DateGenerator(date(2024, 12, 1))

print(f"\n     æ—¥æœŸç”Ÿæˆå™¨æµ‹è¯•:")
daily_dates = generator.generate_daily(5)
print(f"     è¿ç»­5å¤©: {[d.strftime('%m-%d') for d in daily_dates]}")

weekly_dates = generator.generate_weekly(4, 0)  # å‘¨ä¸€
print(f"     è¿ç»­4ä¸ªå‘¨ä¸€: {[d.strftime('%m-%d') for d in weekly_dates]}")

monthly_dates = generator.generate_monthly(3, 15)  # æ¯æœˆ15æ—¥
print(f"     è¿ç»­3ä¸ªæœˆ15æ—¥: {[d.strftime('%m-%d') for d in monthly_dates]}")

workdays = generator.generate_workdays(5)
print(f"     è¿ç»­5ä¸ªå·¥ä½œæ—¥: {[d.strftime('%m-%d') for d in workdays]}")

# 3. æ—¶é—´æ€§èƒ½ä¼˜åŒ–
print(f"\nğŸ“ 3. æ—¶é—´æ€§èƒ½ä¼˜åŒ–:")

# 3.1 æ‰¹é‡æ—¶é—´æ“ä½œä¼˜åŒ–
print(f"\n   âš¡ 3.1 æ‰¹é‡æ—¶é—´æ“ä½œä¼˜åŒ–:")

import time

def slow_date_processing(dates: List[date]) -> List[str]:
    """æ…¢é€Ÿæ—¥æœŸå¤„ç†(é€ä¸ªå¤„ç†)"""
    results = []
    for d in dates:
        # æ¯æ¬¡éƒ½åˆ›å»ºæ–°çš„datetimeå¯¹è±¡
        dt = datetime.combine(d, time.min)
        formatted = dt.strftime('%Y-%m-%d %A')
        results.append(formatted)
    return results

def fast_date_processing(dates: List[date]) -> List[str]:
    """å¿«é€Ÿæ—¥æœŸå¤„ç†(æ‰¹é‡å¤„ç†)"""
    # é¢„ç¼–è¯‘æ ¼å¼åŒ–å­—ç¬¦ä¸²
    format_str = '%Y-%m-%d %A'
    
    # æ‰¹é‡è½¬æ¢
    results = []
    for d in dates:
        # ç›´æ¥ä½¿ç”¨dateå¯¹è±¡çš„æ–¹æ³•
        formatted = d.strftime(format_str)
        results.append(formatted)
    return results

# æ€§èƒ½æµ‹è¯•
test_dates = [date(2024, 1, 1) + timedelta(days=i) for i in range(1000)]

print(f"     æ€§èƒ½æµ‹è¯•(1000ä¸ªæ—¥æœŸ):")

# æ…¢é€Ÿæ–¹æ³•
start_time = time.perf_counter()
slow_results = slow_date_processing(test_dates[:100])
slow_time = time.perf_counter() - start_time

# å¿«é€Ÿæ–¹æ³•
start_time = time.perf_counter()
fast_results = fast_date_processing(test_dates[:100])
fast_time = time.perf_counter() - start_time

print(f"     æ…¢é€Ÿæ–¹æ³•: {slow_time:.6f} ç§’")
print(f"     å¿«é€Ÿæ–¹æ³•: {fast_time:.6f} ç§’")
print(f"     æ€§èƒ½æå‡: {slow_time/fast_time:.2f}x")

# 3.2 ç¼“å­˜ä¼˜åŒ–
print(f"\n   ğŸ’¾ 3.2 ç¼“å­˜ä¼˜åŒ–:")

from functools import lru_cache

@lru_cache(maxsize=128)
def cached_is_workday(year: int, month: int, day: int) -> bool:
    """ç¼“å­˜çš„å·¥ä½œæ—¥åˆ¤æ–­"""
    dt = datetime(year, month, day)
    return dt.weekday() < 5

@lru_cache(maxsize=128)
def cached_format_date(year: int, month: int, day: int, format_type: str = 'default') -> str:
    """ç¼“å­˜çš„æ—¥æœŸæ ¼å¼åŒ–"""
    dt = datetime(year, month, day)
    if format_type == 'iso':
        return dt.isoformat()
    elif format_type == 'short':
        return dt.strftime('%m-%d')
    else:
        return dt.strftime('%Y-%m-%d')

# æµ‹è¯•ç¼“å­˜æ•ˆæœ
print(f"     ç¼“å­˜æµ‹è¯•:")

# ç¬¬ä¸€æ¬¡è°ƒç”¨(æ— ç¼“å­˜)
start_time = time.perf_counter()
for i in range(100):
    cached_is_workday(2024, 12, (i % 28) + 1)
first_call_time = time.perf_counter() - start_time

# ç¬¬äºŒæ¬¡è°ƒç”¨(æœ‰ç¼“å­˜)
start_time = time.perf_counter()
for i in range(100):
    cached_is_workday(2024, 12, (i % 28) + 1)
second_call_time = time.perf_counter() - start_time

print(f"     ç¬¬ä¸€æ¬¡è°ƒç”¨: {first_call_time:.6f} ç§’")
print(f"     ç¬¬äºŒæ¬¡è°ƒç”¨: {second_call_time:.6f} ç§’")
print(f"     ç¼“å­˜æ•ˆæœ: {first_call_time/second_call_time:.2f}x")

# 4. å®é™…åº”ç”¨ç¤ºä¾‹
print(f"\nğŸ“ 4. å®é™…åº”ç”¨ç¤ºä¾‹:")

class ScheduleManager:
    """æ—¥ç¨‹ç®¡ç†å™¨"""
    
    def __init__(self, timezone: str = 'Asia/Shanghai'):
        self.timezone = pytz.timezone(timezone)
        self.holidays = get_chinese_holidays(datetime.now().year)
    
    def add_business_days(self, start_date: date, days: int) -> date:
        """æ·»åŠ è¥ä¸šæ—¥"""
        current_date = start_date
        business_days_added = 0
        
        while business_days_added < days:
            current_date += timedelta(days=1)
            if is_business_day(current_date, self.holidays):
                business_days_added += 1
        
        return current_date
    
    def get_next_meeting_day(self, start_date: date, preferred_weekday: int = None) -> date:
        """è·å–ä¸‹ä¸€ä¸ªä¼šè®®æ—¥"""
        current_date = start_date + timedelta(days=1)
        
        while True:
            # æ£€æŸ¥æ˜¯å¦ä¸ºè¥ä¸šæ—¥
            if is_business_day(current_date, self.holidays):
                # å¦‚æœæŒ‡å®šäº†é¦–é€‰æ˜ŸæœŸå‡ 
                if preferred_weekday is None or current_date.weekday() == preferred_weekday:
                    return current_date
            current_date += timedelta(days=1)
    
    def generate_report_schedule(self, start_date: date, end_date: date, 
                                 frequency: str = 'monthly') -> List[date]:
        """ç”ŸæˆæŠ¥å‘Šè®¡åˆ’"""
        schedule = []
        current_date = start_date
        
        if frequency == 'daily':
            while current_date <= end_date:
                if is_business_day(current_date, self.holidays):
                    schedule.append(current_date)
                current_date += timedelta(days=1)
        
        elif frequency == 'weekly':
            while current_date <= end_date:
                if current_date.weekday() == 0:  # å‘¨ä¸€
                    if is_business_day(current_date, self.holidays):
                        schedule.append(current_date)
                current_date += timedelta(days=1)
        
        elif frequency == 'monthly':
            while current_date <= end_date:
                # æ¯æœˆæœ€åä¸€ä¸ªå·¥ä½œæ—¥
                if current_date.day == 1:  # æœˆåˆ
                    # æ‰¾åˆ°ä¸Šä¸ªæœˆçš„æœ€åä¸€å¤©
                    last_day = current_date - timedelta(days=1)
                    if is_business_day(last_day, self.holidays):
                        schedule.append(last_day)
                current_date += timedelta(days=1)
        
        return schedule
    
    def format_schedule(self, dates: List[date], format_type: str = 'default') -> List[str]:
        """æ ¼å¼åŒ–æ—¥ç¨‹"""
        formatted_dates = []
        for d in dates:
            dt = datetime.combine(d, time.min, self.timezone)
            
            if format_type == 'detailed':
                formatted = dt.strftime('%Yå¹´%mæœˆ%dæ—¥ %A')
            elif format_type == 'short':
                formatted = dt.strftime('%m-%d')
            else:
                formatted = dt.strftime('%Y-%m-%d')
            
            formatted_dates.append(formatted)
        
        return formatted_dates

# æµ‹è¯•æ—¥ç¨‹ç®¡ç†å™¨
print(f"\n   æµ‹è¯•æ—¥ç¨‹ç®¡ç†å™¨:")

schedule_manager = ScheduleManager()
today = date.today()

# æ·»åŠ è¥ä¸šæ—¥
future_date = schedule_manager.add_business_days(today, 10)
print(f"\n     ä»ä»Šå¤©èµ·10ä¸ªè¥ä¸šæ—¥å: {future_date}")

# è·å–ä¸‹ä¸€ä¸ªä¼šè®®æ—¥(å‘¨äº”)
next_meeting = schedule_manager.get_next_meeting_day(today, preferred_weekday=4)
print(f"     ä¸‹ä¸€ä¸ªå‘¨äº”ä¼šè®®æ—¥: {next_meeting}")

# ç”ŸæˆæŠ¥å‘Šè®¡åˆ’
start_date = date(2024, 12, 1)
end_date = date(2024, 12, 31)

daily_schedule = schedule_manager.generate_report_schedule(start_date, end_date, 'daily')
weekly_schedule = schedule_manager.generate_report_schedule(start_date, end_date, 'weekly')
monthly_schedule = schedule_manager.generate_report_schedule(start_date, end_date, 'monthly')

print(f"\n     2024å¹´12æœˆæŠ¥å‘Šè®¡åˆ’:")
print(f"     æ—¥æŠ¥: {len(daily_schedule)} æ¬¡ (å‰5æ¬¡: {schedule_manager.format_schedule(daily_schedule[:5], 'short')})")
print(f"     å‘¨æŠ¥: {len(weekly_schedule)} æ¬¡ (å…¨éƒ¨: {schedule_manager.format_schedule(weekly_schedule, 'short')})")
print(f"     æœˆæŠ¥: {len(monthly_schedule)} æ¬¡ (å…¨éƒ¨: {schedule_manager.format_schedule(monthly_schedule, 'short')})")

print(f"\nâœ… é«˜çº§æ—¥æœŸæ—¶é—´æ“ä½œå®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å·¥ä½œæ—¥å’ŒèŠ‚å‡æ—¥è®¡ç®—")
print(f"   âœ“ ç†è§£å‘¨æœŸæ€§æ—¶é—´å¤„ç†")
print(f"   âœ“ èƒ½è¿›è¡Œæ—¶é—´æ€§èƒ½ä¼˜åŒ–")
print(f"   âœ“ æŒæ¡å¤æ‚çš„æ—¶é—´åœºæ™¯å¤„ç†")
````
