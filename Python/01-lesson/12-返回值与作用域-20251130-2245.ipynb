````python
print("ğŸ”„ å‡½æ•°è¿”å›å€¼è¯¦è§£:")
print("=" * 50)

# å•ä¸ªè¿”å›å€¼
print(f"ğŸ“ 1. å•ä¸ªè¿”å›å€¼:")
def calculate_area(length, width):
    """è®¡ç®—çŸ©å½¢é¢ç§¯"""
    area = length * width
    return area

result = calculate_area(10, 5)
print(f"   çŸ©å½¢é¢ç§¯: {result}")

# å¤šä¸ªè¿”å›å€¼ï¼ˆå…ƒç»„ï¼‰
print(f"\nğŸ“ 2. å¤šä¸ªè¿”å›å€¼ï¼ˆå…ƒç»„ï¼‰:")
def calculate_rectangle_properties(length, width):
    """è®¡ç®—çŸ©å½¢çš„å¤šä¸ªå±æ€§"""
    area = length * width
    perimeter = 2 * (length + width)
    diagonal = (length ** 2 + width ** 2) ** 0.5
    return area, perimeter, diagonal

# æ¥æ”¶å¤šä¸ªè¿”å›å€¼
area, perimeter, diagonal = calculate_rectangle_properties(8, 6)
print(f"   çŸ©å½¢å±æ€§ (8x6):")
print(f"     é¢ç§¯: {area}")
print(f"     å‘¨é•¿: {perimeter}")
print(f"     å¯¹è§’çº¿: {diagonal:.2f}")

# ä½œä¸ºå…ƒç»„æ¥æ”¶
properties = calculate_rectangle_properties(12, 5)
print(f"\n   ä½œä¸ºå…ƒç»„æ¥æ”¶: {properties}")
print(f"   ç±»å‹: {type(properties)}")

# åˆ—è¡¨è¿”å›å€¼
print(f"\nğŸ“ 3. åˆ—è¡¨è¿”å›å€¼:")
def generate_fibonacci(n):
    """ç”Ÿæˆæ–æ³¢é‚£å¥‘æ•°åˆ—"""
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib

fib_sequence = generate_fibonacci(10)
print(f"   æ–æ³¢é‚£å¥‘æ•°åˆ—(å‰10é¡¹): {fib_sequence}")

# å­—å…¸è¿”å›å€¼
print(f"\nğŸ“ 4. å­—å…¸è¿”å›å€¼:")
def analyze_string(text):
    """åˆ†æå­—ç¬¦ä¸²å¹¶è¿”å›ç»Ÿè®¡ä¿¡æ¯"""
    if not text:
        return {"error": "ç©ºå­—ç¬¦ä¸²"}
    
    words = text.split()
    characters = len(text)
    words_count = len(words)
    
    # ç»Ÿè®¡å­—ç¬¦ç±»å‹
    letters = sum(1 for c in text if c.isalpha())
    digits = sum(1 for c in text if c.isdigit())
    spaces = sum(1 for c in text if c.isspace())
    
    return {
        "length": characters,
        "words": words_count,
        "letters": letters,
        "digits": digits,
        "spaces": spaces,
        "word_list": words[:5]  # è¿”å›å‰5ä¸ªå•è¯
    }

analysis = analyze_string("Hello Python 123! This is a test.")
print(f"   å­—ç¬¦ä¸²åˆ†æç»“æœ:")
for key, value in analysis.items():
    print(f"     {key}: {value}")

# æ¡ä»¶è¿”å›å€¼
print(f"\nğŸ“ 5. æ¡ä»¶è¿”å›å€¼:")
def get_grade_info(score):
    """æ ¹æ®åˆ†æ•°è¿”å›ç­‰çº§ä¿¡æ¯"""
    if score < 0 or score > 100:
        return {"error": "åˆ†æ•°å¿…é¡»åœ¨0-100ä¹‹é—´"}
    
    if score >= 90:
        grade = "A"
        comment = "ä¼˜ç§€"
    elif score >= 80:
        grade = "B"
        comment = "è‰¯å¥½"
    elif score >= 70:
        grade = "C"
        comment = "ä¸­ç­‰"
    elif score >= 60:
        grade = "D"
        comment = "åŠæ ¼"
    else:
        grade = "F"
        comment = "ä¸åŠæ ¼"
    
    return {
        "score": score,
        "grade": grade,
        "comment": comment,
        "passed": score >= 60
    }

test_scores = [95, 85, 75, 65, 55, 105]
print(f"   åˆ†æ•°ç­‰çº§åˆ†æ:")
for score in test_scores:
    result = get_grade_info(score)
    if "error" in result:
        print(f"     åˆ†æ•°{score}: {result['error']}")
    else:
        print(f"     åˆ†æ•°{score}: {result['grade']}çº§ ({result['comment']})")

# æ—©æœŸè¿”å›æ¨¡å¼
print(f"\nğŸ“ 6. æ—©æœŸè¿”å›æ¨¡å¼:")
def validate_and_process(data):
    """éªŒè¯å¹¶å¤„ç†æ•°æ®"""
    # è¾“å…¥éªŒè¯
    if not data:
        return None, "æ•°æ®ä¸èƒ½ä¸ºç©º"
    
    if not isinstance(data, list):
        return None, "æ•°æ®å¿…é¡»æ˜¯åˆ—è¡¨"
    
    if len(data) == 0:
        return None, "åˆ—è¡¨ä¸èƒ½ä¸ºç©º"
    
    # æ•°æ®å¤„ç†
    try:
        # è¿‡æ»¤æ•°å­—
        numbers = [x for x in data if isinstance(x, (int, float))]
        
        if not numbers:
            return None, "åˆ—è¡¨ä¸­æ²¡æœ‰æ•°å­—"
        
        # è®¡ç®—ç»Ÿè®¡ä¿¡æ¯
        result = {
            "count": len(numbers),
            "sum": sum(numbers),
            "average": sum(numbers) / len(numbers),
            "max": max(numbers),
            "min": min(numbers)
        }
        
        return result, "å¤„ç†æˆåŠŸ"
    
    except Exception as e:
        return None, f"å¤„ç†é”™è¯¯: {str(e)}"

# æµ‹è¯•æ—©æœŸè¿”å›
test_data_sets = [
    [],
    "not a list",
    ["a", "b", "c"],
    [1, 2, 3, "hello", 4.5],
    [10, 20, 30, 40, 50]
]

print(f"   æ•°æ®éªŒè¯å’Œå¤„ç†æµ‹è¯•:")
for i, data in enumerate(test_data_sets, 1):
    result, message = validate_and_process(data)
    print(f"\n     æµ‹è¯•{i}: {data}")
    if result:
        print(f"       ç»“æœ: {result}")
    print(f"       æ¶ˆæ¯: {message}")

print(f"\nâœ… å‡½æ•°è¿”å›å€¼è¯¦è§£å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å„ç§è¿”å›å€¼ç±»å‹ä½¿ç”¨")
print(f"   âœ“ ç†è§£æ¡ä»¶è¿”å›å’Œæ—©æœŸè¿”å›")
print(f"   âœ“ èƒ½è®¾è®¡åˆç†çš„å‡½æ•°è¿”å›å€¼")
````

### ä½œç”¨åŸŸè§„åˆ™è¯¦è§£ [â­åŸºç¡€]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šä½œç”¨åŸŸå†³å®šäº†å˜é‡çš„å¯è®¿é—®èŒƒå›´ï¼ŒPythoné‡‡ç”¨LEGBè§„åˆ™ï¼ˆLocal, Enclosing, Global, Built-inï¼‰è¿›è¡Œå˜é‡æŸ¥æ‰¾ã€‚

**å­¦ä¹ è¦æ±‚**ï¼š
- æŒæ¡LEGBä½œç”¨åŸŸè§„åˆ™
- ç†è§£å˜é‡ç”Ÿå‘½å‘¨æœŸå’Œå†…å­˜ç®¡ç†
- èƒ½æ­£ç¡®å¤„ç†ä½œç”¨åŸŸç›¸å…³é—®é¢˜

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°ä¸åŒä½œç”¨åŸŸçš„å˜é‡å¤„ç†
- è¿›è¡Œä½œç”¨åŸŸé“¾éªŒè¯
- åº”ç”¨ä½œç”¨åŸŸçŸ¥è¯†è§£å†³å®é™…é—®é¢˜
- éªŒè¯ç‚¹ï¼šèƒ½æ­£ç¡®ç†è§£å’Œåº”ç”¨ä½œç”¨åŸŸè§„åˆ™

````python
print("ğŸŒ ä½œç”¨åŸŸè§„åˆ™è¯¦è§£:")
print("=" * 50)

# LEGBè§„åˆ™æ¼”ç¤º
print(f"ğŸ“ 1. LEGBè§„åˆ™æ¼”ç¤º:")

# Built-inä½œç”¨åŸŸï¼ˆå†…ç½®å‡½æ•°ï¼‰
print(f"   Built-inä½œç”¨åŸŸ: len(), sum(), max()ç­‰å†…ç½®å‡½æ•°")

# Globalä½œç”¨åŸŸ
global_var = "å…¨å±€å˜é‡"

def demonstrate_scope():
    """æ¼”ç¤ºä½œç”¨åŸŸè§„åˆ™"""
    # Localä½œç”¨åŸŸ
    local_var = "å±€éƒ¨å˜é‡"
    
    print(f"   Local: {local_var}")
    print(f"   Global: {global_var}")
    print(f"   Built-in: {len('test')}")

demonstrate_scope()

# Enclosingä½œç”¨åŸŸï¼ˆé—­åŒ…ï¼‰
print(f"\nğŸ“ 2. Enclosingä½œç”¨åŸŸï¼ˆé—­åŒ…ï¼‰:")

def outer_function():
    """å¤–å±‚å‡½æ•°"""
    enclosing_var = "å¤–å±‚å‡½æ•°å˜é‡"
    
    def inner_function():
        """å†…å±‚å‡½æ•°"""
        local_var = "å†…å±‚å‡½æ•°å˜é‡"
        print(f"     Local: {local_var}")
        print(f"     Enclosing: {enclosing_var}")
        print(f"     Global: {global_var}")
    
    print(f"   åœ¨å¤–å±‚å‡½æ•°ä¸­:")
    print(f"     Local: {enclosing_var}")
    print(f"     Global: {global_var}")
    
    inner_function()

outer_function()

# å˜é‡é®è”½
print(f"\nğŸ“ 3. å˜é‡é®è”½:")

x = 10  # Globalå˜é‡

def shadowing_demo():
    """æ¼”ç¤ºå˜é‡é®è”½"""
    x = 20  # Localå˜é‡ï¼Œé®è”½äº†Globalå˜é‡
    print(f"   å‡½æ•°å†…éƒ¨x: {x}")
    
    def inner():
        x = 30  # æ›´å†…å±‚çš„Localå˜é‡
        print(f"     å†…å±‚å‡½æ•°x: {x}")
    
    inner()
    print(f"   å›åˆ°å¤–å±‚å‡½æ•°x: {x}")

print(f"   è°ƒç”¨å‰å…¨å±€x: {x}")
shadowing_demo()
print(f"   è°ƒç”¨åå…¨å±€x: {x}")

# globalå…³é”®å­—
print(f"\nğŸ“ 4. globalå…³é”®å­—:")

counter = 0

def increment_counter():
    """å¢åŠ å…¨å±€è®¡æ•°å™¨"""
    global counter  # å£°æ˜è¦ä¿®æ”¹å…¨å±€å˜é‡
    counter += 1
    print(f"   è®¡æ•°å™¨å¢åŠ åˆ°: {counter}")

def reset_counter():
    """é‡ç½®å…¨å±€è®¡æ•°å™¨"""
    global counter
    counter = 0
    print(f"   è®¡æ•°å™¨é‡ç½®ä¸º: {counter}")

print(f"   åˆå§‹è®¡æ•°å™¨: {counter}")
increment_counter()
increment_counter()
increment_counter()
reset_counter()
increment_counter()

# nonlocalå…³é”®å­—
print(f"\nğŸ“ 5. nonlocalå…³é”®å­—:")

def create_counter():
    """åˆ›å»ºè®¡æ•°å™¨é—­åŒ…"""
    count = 0  # å¤–å±‚å‡½æ•°çš„å±€éƒ¨å˜é‡
    
    def increment():
        nonlocal count  # å£°æ˜ä½¿ç”¨å¤–å±‚å‡½æ•°çš„å˜é‡
        count += 1
        return count
    
    def decrement():
        nonlocal count
        count -= 1
        return count
    
    def get_count():
        return count
    
    def reset():
        nonlocal count
        count = 0
        return count
    
    return increment, decrement, get_count, reset

# åˆ›å»ºè®¡æ•°å™¨
inc, dec, get, reset = create_counter()

print(f"   é—­åŒ…è®¡æ•°å™¨æµ‹è¯•:")
print(f"     åˆå§‹å€¼: {get()}")
print(f"     å¢åŠ : {inc()}")
print(f"     å¢åŠ : {inc()}")
print(f"     å‡å°‘: {dec()}")
print(f"     å½“å‰å€¼: {get()}")
print(f"     é‡ç½®: {reset()}")
print(f"     é‡ç½®å: {get()}")

# ä½œç”¨åŸŸçš„å®é™…åº”ç”¨
print(f"\nğŸ“ 6. ä½œç”¨åŸŸå®é™…åº”ç”¨ - é…ç½®ç®¡ç†:")

class ConfigManager:
    """é…ç½®ç®¡ç†å™¨ç±»"""
    def __init__(self):
        self._config = {}
    
    def set(self, key, value):
        self._config[key] = value
    
    def get(self, key, default=None):
        return self._config.get(key, default)
    
    def update(self, new_config):
        self._config.update(new_config)

# å…¨å±€é…ç½®å®ä¾‹
app_config = ConfigManager()
app_config.set("debug", False)
app_config.set("version", "1.0.0")

def debug_log(message):
    """æ ¹æ®é…ç½®è¾“å‡ºè°ƒè¯•ä¿¡æ¯"""
    global app_config
    if app_config.get("debug", False):
        print(f"[DEBUG] {message}")

def set_debug_mode(enabled):
    """è®¾ç½®è°ƒè¯•æ¨¡å¼"""
    global app_config
    app_config.set("debug", enabled)
    status = "å¼€å¯" if enabled else "å…³é—­"
    print(f"   è°ƒè¯•æ¨¡å¼å·²{status}")

print(f"   é…ç½®ç®¡ç†æµ‹è¯•:")
debug_log("è¿™æ¡æ¶ˆæ¯ä¸ä¼šæ˜¾ç¤º")

set_debug_mode(True)
debug_log("è¿™æ¡æ¶ˆæ¯ä¼šæ˜¾ç¤º")
debug_log("è°ƒè¯•ä¿¡æ¯å¾ˆé‡è¦")

set_debug_mode(False)
debug_log("è¿™æ¡æ¶ˆæ¯åˆä¸ä¼šæ˜¾ç¤ºäº†")

# æ•°æ®å¤„ç†ç®¡é“ä¸­çš„ä½œç”¨åŸŸ
print(f"\nğŸ“ 7. æ•°æ®å¤„ç†ç®¡é“ä¸­çš„ä½œç”¨åŸŸ:")

def create_data_processor(initial_data):
    """åˆ›å»ºæ•°æ®å¤„ç†ç®¡é“"""
    data = initial_data.copy()
    operations_log = []
    
    def add_item(item):
        nonlocal data, operations_log
        data.append(item)
        operations_log.append(f"æ·»åŠ : {item}")
    
    def remove_item(index):
        nonlocal data, operations_log
        if 0 <= index < len(data):
            removed = data.pop(index)
            operations_log.append(f"åˆ é™¤ç´¢å¼•{index}: {removed}")
            return removed
        return None
    
    def filter_items(predicate):
        nonlocal data, operations_log
        original_count = len(data)
        data = [item for item in data if predicate(item)]
        removed_count = original_count - len(data)
        operations_log.append(f"è¿‡æ»¤: ç§»é™¤{removed_count}é¡¹")
    
    def get_data():
        return data.copy()
    
    def get_log():
        return operations_log.copy()
    
    def reset(new_data):
        nonlocal data, operations_log
        data = new_data.copy()
        operations_log.clear()
    
    return {
        "add": add_item,
        "remove": remove_item,
        "filter": filter_items,
        "get_data": get_data,
        "get_log": get_log,
        "reset": reset
    }

# åˆ›å»ºæ•°æ®å¤„ç†å™¨
processor = create_data_processor([1, 2, 3, 4, 5])

print(f"   æ•°æ®å¤„ç†ç®¡é“æµ‹è¯•:")
print(f"     åˆå§‹æ•°æ®: {processor['get_data']()}")

processor["add"](6)
processor["add"](7)
print(f"     æ·»åŠ å: {processor['get_data']()}")

processor["remove"](2)
print(f"     åˆ é™¤ç´¢å¼•2å: {processor['get_data']()}")

processor["filter"](lambda x: x % 2 == 0)
print(f"     è¿‡æ»¤å¶æ•°å: {processor['get_data']()}")

print(f"     æ“ä½œæ—¥å¿—: {processor['get_log']()}")

print(f"\nâœ… ä½œç”¨åŸŸè§„åˆ™è¯¦è§£å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡LEGBä½œç”¨åŸŸè§„åˆ™")
print(f"   âœ“ ç†è§£å˜é‡ç”Ÿå‘½å‘¨æœŸ")
print(f"   âœ“ èƒ½æ­£ç¡®å¤„ç†ä½œç”¨åŸŸé—®é¢˜")
````

````python
print("ğŸ”— é—­åŒ…ä¸ä½œç”¨åŸŸé“¾:")
print("=" * 50)

# åŸºæœ¬é—­åŒ…
print(f"ğŸ“ 1. åŸºæœ¬é—­åŒ…:")

def outer_func(x):
    """å¤–å±‚å‡½æ•°"""
    def inner_func(y):
        """å†…å±‚å‡½æ•°ï¼Œå½¢æˆé—­åŒ…"""
        return x + y  # è®¿é—®å¤–éƒ¨å˜é‡x
    return inner_func

# åˆ›å»ºé—­åŒ…
add_five = outer_func(5)
add_ten = outer_func(10)

print(f"   add_five(3) = {add_five(3)}")
print(f"   add_five(7) = {add_five(7)}")
print(f"   add_ten(3) = {add_ten(3)}")
print(f"   add_ten(7) = {add_ten(7)}")

# é—­åŒ…æ•è·å˜é‡
print(f"\nğŸ“ 2. é—­åŒ…æ•è·å˜é‡:")

def make_multiplier(factor):
    """åˆ›å»ºä¹˜æ³•å™¨é—­åŒ…"""
    def multiplier(number):
        return number * factor
    return multiplier

double = make_multiplier(2)
triple = make_multiplier(3)
half = make_multiplier(0.5)

print(f"   double(10) = {double(10)}")
print(f"   triple(10) = {triple(10)}")
print(f"   half(10) = {half(10)}")

# é—­åŒ…ä¸å¾ªç¯
print(f"\nğŸ“ 3. é—­åŒ…ä¸å¾ªç¯ï¼ˆå¸¸è§é™·é˜±ï¼‰:")

# é”™è¯¯æ–¹å¼ï¼šæ‰€æœ‰é—­åŒ…éƒ½æ•è·åŒä¸€ä¸ªå˜é‡çš„æœ€ç»ˆå€¼
def create_multipliers_wrong():
    multipliers = []
    for i in range(1, 4):
        def multiplier(x):
            return x * i  # æ•è·å˜é‡iï¼Œä½†iä¼šå˜åŒ–
        multipliers.append(multiplier)
    return multipliers

# æ­£ç¡®æ–¹å¼ï¼šä½¿ç”¨é»˜è®¤å‚æ•°æ•è·å½“å‰å€¼
def create_multipliers_correct():
    multipliers = []
    for i in range(1, 4):
        def multiplier(x, factor=i):  # ä½¿ç”¨é»˜è®¤å‚æ•°æ•è·å½“å‰iå€¼
            return x * factor
        multipliers.append(multiplier)
    return multipliers

print(f"   é”™è¯¯æ–¹å¼ç»“æœ:")
wrong_multipliers = create_multipliers_wrong()
for i, mul in enumerate(wrong_multipliers, 1):
    print(f"     ä¹˜æ³•å™¨{i}(10) = {mul(10)}")

print(f"\n   æ­£ç¡®æ–¹å¼ç»“æœ:")
correct_multipliers = create_multipliers_correct()
for i, mul in enumerate(correct_multipliers, 1):
    print(f"     ä¹˜æ³•å™¨{i}(10) = {mul(10)}")

# é—­åŒ…ä½œä¸ºæ•°æ®å°è£…
print(f"\nğŸ“ 4. é—­åŒ…ä½œä¸ºæ•°æ®å°è£…:")

def create_counter():
    """åˆ›å»ºè®¡æ•°å™¨é—­åŒ…"""
    count = 0  # ç§æœ‰å˜é‡
    
    def increment():
        nonlocal count
        count += 1
        return count
    
    def decrement():
        nonlocal count
        count -= 1
        return count
    
    def get_count():
        return count
    
    def reset():
        nonlocal count
        count = 0
        return count
    
    # è¿”å›å­—å…¸åŒ…å«æ‰€æœ‰æ“ä½œ
    return {
        'increment': increment,
        'decrement': decrement,
        'get_count': get_count,
        'reset': reset
    }

# åˆ›å»ºè®¡æ•°å™¨å®ä¾‹
counter1 = create_counter()
counter2 = create_counter()

print(f"   è®¡æ•°å™¨1æµ‹è¯•:")
print(f"     åˆå§‹: {counter1['get_count']()}")
print(f"     å¢åŠ : {counter1['increment']()}")
print(f"     å¢åŠ : {counter1['increment']()}")
print(f"     å‡å°‘: {counter1['decrement']()}")
print(f"     å½“å‰: {counter1['get_count']()}")

print(f"\n   è®¡æ•°å™¨2æµ‹è¯•ï¼ˆç‹¬ç«‹ï¼‰:")
print(f"     åˆå§‹: {counter2['get_count']()}")
print(f"     å¢åŠ : {counter2['increment']()}")
print(f"     é‡ç½®: {counter2['reset']()}")
print(f"     é‡ç½®å: {counter2['get_count']()}")

# é—­åŒ…å®ç°ç¼“å­˜
print(f"\nğŸ“ 5. é—­åŒ…å®ç°ç¼“å­˜:")

def memoize(func):
    """ç®€å•çš„è®°å¿†åŒ–è£…é¥°å™¨ï¼ˆä½¿ç”¨é—­åŒ…ï¼‰"""
    cache = {}  # ç¼“å­˜å­—å…¸
    
    def memoized_func(*args):
        if args in cache:
            print(f"     ç¼“å­˜å‘½ä¸­: {args}")
            return cache[args]
        
        result = func(*args)
        cache[args] = result
        print(f"     è®¡ç®—å¹¶ç¼“å­˜: {args} -> {result}")
        return result
    
    memoized_func.cache = cache  # æš´éœ²ç¼“å­˜
    return memoized_func

# å®šä¹‰ä¸€ä¸ªè®¡ç®—å¯†é›†çš„å‡½æ•°
def fibonacci(n):
    """è®¡ç®—æ–æ³¢é‚£å¥‘æ•°ï¼ˆé€’å½’ç‰ˆæœ¬ï¼‰"""
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# åˆ›å»ºè®°å¿†åŒ–ç‰ˆæœ¬
memo_fib = memoize(fibonacci)

print(f"   è®°å¿†åŒ–æ–æ³¢é‚£å¥‘æµ‹è¯•:")
print(f"     fibonacci(5): {memo_fib(5)}")
print(f"     fibonacci(5): {memo_fib(5)}")  # ä»ç¼“å­˜è·å–
print(f"     fibonacci(6): {memo_fib(6)}")
print(f"     fibonacci(6): {memo_fib(6)}")  # ä»ç¼“å­˜è·å–

# é—­åŒ…å®ç°å·¥å‚å‡½æ•°
print(f"\nğŸ“ 6. é—­åŒ…å®ç°å·¥å‚å‡½æ•°:")

def create_validator(min_value=None, max_value=None, data_type=None):
    """åˆ›å»ºéªŒè¯å™¨å·¥å‚å‡½æ•°"""
    def validator(value):
        errors = []
        
        # ç±»å‹æ£€æŸ¥
        if data_type and not isinstance(value, data_type):
            errors.append(f"ç±»å‹é”™è¯¯: æœŸæœ›{data_type.__name__}, å®é™…{type(value).__name__}")
        
        # æœ€å°å€¼æ£€æŸ¥
        if min_value is not None and value < min_value:
            errors.append(f"å€¼å¤ªå°: æœ€å°å€¼ä¸º{min_value}")
        
        # æœ€å¤§å€¼æ£€æŸ¥
        if max_value is not None and value > max_value:
            errors.append(f"å€¼å¤ªå¤§: æœ€å¤§å€¼ä¸º{max_value}")
        
        return len(errors) == 0, errors
    
    return validator

# åˆ›å»ºå„ç§éªŒè¯å™¨
age_validator = create_validator(min_value=0, max_value=150, data_type=int)
score_validator = create_validator(min_value=0, max_value=100, data_type=(int, float))
name_validator = create_validator(data_type=str)

# æµ‹è¯•éªŒè¯å™¨
test_values = [
    ("å¹´é¾„éªŒè¯å™¨", age_validator, [25, -5, 200, "30"]),
    ("åˆ†æ•°éªŒè¯å™¨", score_validator, [85, -10, 105, 95.5]),
    ("å§“åéªŒè¯å™¨", name_validator, ["å¼ ä¸‰", 123, None])
]

print(f"   éªŒè¯å™¨å·¥å‚æµ‹è¯•:")
for validator_name, validator, values in test_values:
    print(f"\n     {validator_name}:")
    for value in values:
        is_valid, errors = validator(value)
        status = "âœ…" if is_valid else "âŒ"
        print(f"       {status} {value}: {', '.join(errors) if errors else 'æœ‰æ•ˆ'}")

# é—­åŒ…å®ç°çŠ¶æ€ç®¡ç†
print(f"\nğŸ“ 7. é—­åŒ…å®ç°çŠ¶æ€ç®¡ç†:")

def create_state_manager(initial_state=None):
    """åˆ›å»ºçŠ¶æ€ç®¡ç†å™¨"""
    state = initial_state.copy() if initial_state else {}
    history = []  # çŠ¶æ€å†å²
    
    def get_state(key=None):
        if key is None:
            return state.copy()
        return state.get(key)
    
    def set_state(**updates):
        nonlocal state
        # ä¿å­˜å½“å‰çŠ¶æ€åˆ°å†å²
        history.append(state.copy())
        
        # æ›´æ–°çŠ¶æ€
        state.update(updates)
        return state.copy()
    
    def reset_state(new_state=None):
        nonlocal state, history
        history.append(state.copy())
        state = new_state.copy() if new_state else {}
        return state.copy()
    
    def get_history():
        return history.copy()
    
    def undo():
        nonlocal state
        if history:
            state = history.pop()
        return state.copy()
    
    return {
        'get': get_state,
        'set': set_state,
        'reset': reset_state,
        'history': get_history,
        'undo': undo
    }

# åˆ›å»ºçŠ¶æ€ç®¡ç†å™¨
state_mgr = create_state_manager({'count': 0, 'name': 'åˆå§‹çŠ¶æ€'})

print(f"   çŠ¶æ€ç®¡ç†å™¨æµ‹è¯•:")
print(f"     åˆå§‹çŠ¶æ€: {state_mgr['get']()}")

state_mgr['set'](count=5, name='æ›´æ–°çŠ¶æ€')
print(f"     æ›´æ–°å: {state_mgr['get']()}")

state_mgr['set'](count=10, status='active')
print(f"     å†æ¬¡æ›´æ–°: {state_mgr['get']()}")

print(f"     çŠ¶æ€å†å²: {state_mgr['history']()}")

state_mgr['undo']()
print(f"     æ’¤é”€å: {state_mgr['get']()}")

state_mgr['undo']()
print(f"     å†æ¬¡æ’¤é”€: {state_mgr['get']()}")

print(f"\nâœ… é—­åŒ…ä¸ä½œç”¨åŸŸé“¾å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡é—­åŒ…åˆ›å»ºå’Œä½¿ç”¨")
print(f"   âœ“ ç†è§£ä½œç”¨åŸŸé“¾å·¥ä½œåŸç†")
print(f"   âœ“ èƒ½åº”ç”¨é—­åŒ…è§£å†³å®é™…é—®é¢˜")
````

````python
print("ğŸ’¼ å®é™…åº”ç”¨æ¡ˆä¾‹:")
print("=" * 50)

# æ¡ˆä¾‹1ï¼šæ™ºèƒ½ç¼“å­˜ç³»ç»Ÿ
print(f"ğŸ“ 1. æ™ºèƒ½ç¼“å­˜ç³»ç»Ÿ:")

def create_smart_cache(max_size=100, ttl=None):
    """åˆ›å»ºæ™ºèƒ½ç¼“å­˜ç³»ç»Ÿ"""
    import time
    
    cache = {}
    access_times = {}  # è®°å½•è®¿é—®æ—¶é—´
    creation_times = {}  # è®°å½•åˆ›å»ºæ—¶é—´
    
    def is_expired(key):
        """æ£€æŸ¥ç¼“å­˜æ˜¯å¦è¿‡æœŸ"""
        if ttl is None:
            return False
        return time.time() - creation_times.get(key, 0) > ttl
    
    def evict_lru():
        """é©±é€æœ€è¿‘æœ€å°‘ä½¿ç”¨çš„é¡¹"""
        if len(cache) >= max_size:
            # æ‰¾åˆ°æœ€ä¹…æœªè®¿é—®çš„é”®
            lru_key = min(access_times.keys(), key=lambda k: access_times[k])
            del cache[lru_key]
            del access_times[lru_key]
            del creation_times[lru_key]
            print(f"     é©±é€LRUé¡¹: {lru_key}")
    
    def get(key):
        """è·å–ç¼“å­˜å€¼"""
        if key not in cache:
            return None
        
        if is_expired(key):
            del cache[key]
            del access_times[key]
            del creation_times[key]
            print(f"     ç¼“å­˜è¿‡æœŸ: {key}")
            return None
        
        access_times[key] = time.time()
        return cache[key]
    
    def set(key, value):
        """è®¾ç½®ç¼“å­˜å€¼"""
        current_time = time.time()
        
        # å¦‚æœé”®å·²å­˜åœ¨ï¼Œæ›´æ–°æ—¶é—´
        if key in cache:
            access_times[key] = current_time
            creation_times[key] = current_time
            cache[key] = value
            return
        
        # é©±é€è¿‡æœŸé¡¹
        evict_lru()
        
        # æ·»åŠ æ–°é¡¹
        cache[key] = value
        access_times[key] = current_time
        creation_times[key] = current_time
        print(f"     ç¼“å­˜æ–°å¢: {key}")
    
    def clear():
        """æ¸…ç©ºç¼“å­˜"""
        nonlocal cache, access_times, creation_times
        cache.clear()
        access_times.clear()
        creation_times.clear()
    
    def get_stats():
        """è·å–ç¼“å­˜ç»Ÿè®¡"""
        return {
            "size": len(cache),
            "max_size": max_size,
            "ttl": ttl,
            "keys": list(cache.keys())
        }
    
    return {
        'get': get,
        'set': set,
        'clear': clear,
        'stats': get_stats
    }

# åˆ›å»ºå¹¶æµ‹è¯•ç¼“å­˜ç³»ç»Ÿ
cache = create_smart_cache(max_size=3, ttl=5)  # æœ€å¤š3é¡¹ï¼Œ5ç§’è¿‡æœŸ

print(f"   æ™ºèƒ½ç¼“å­˜æµ‹è¯•:")
cache['set']('key1', 'value1')
cache['set']('key2', 'value2')
cache['set']('key3', 'value3')
print(f"     ç¼“å­˜çŠ¶æ€: {cache['stats']()}")

print(f"     è·å–key1: {cache['get']('key1')}")
print(f"     è·å–key4: {cache['get']('key4')}")

# æ·»åŠ ç¬¬4é¡¹ï¼Œä¼šè§¦å‘LRUé©±é€
cache['set']('key4', 'value4')
print(f"     æ·»åŠ key4å: {cache['stats']()}")

# æ¡ˆä¾‹2ï¼šå‡½æ•°å·¥å‚å’Œé…ç½®ç®¡ç†
print(f"\nğŸ“ 2. å‡½æ•°å·¥å‚å’Œé…ç½®ç®¡ç†:")

def create_api_client(base_url, default_headers=None, timeout=30):
    """åˆ›å»ºAPIå®¢æˆ·ç«¯å·¥å‚å‡½æ•°"""
    import json
    
    # é…ç½®çŠ¶æ€
    config = {
        'base_url': base_url,
        'headers': default_headers or {},
        'timeout': timeout,
        'last_request': None,
        'request_count': 0
    }
    
    def make_request(endpoint, method='GET', data=None, headers=None):
        """å‘èµ·APIè¯·æ±‚ï¼ˆæ¨¡æ‹Ÿï¼‰"""
        # åˆå¹¶headers
        request_headers = config['headers'].copy()
        if headers:
            request_headers.update(headers)
        
        # æ„å»ºå®Œæ•´URL
        url = f"{config['base_url']}/{endpoint.lstrip('/')}"
        
        # æ›´æ–°ç»Ÿè®¡
        config['request_count'] += 1
        config['last_request'] = {
            'url': url,
            'method': method,
            'headers': request_headers,
            'data': data
        }
        
        # æ¨¡æ‹Ÿè¯·æ±‚å“åº”
        response = {
            'url': url,
            'method': method,
            'status': 200,
            'data': data,
            'response_data': f"Mock response for {method} {endpoint}"
        }
        
        return response
    
    def update_config(**kwargs):
        """æ›´æ–°é…ç½®"""
        config.update(kwargs)
    
    def get_config():
        """è·å–å½“å‰é…ç½®"""
        return config.copy()
    
    def get_stats():
        """è·å–ä½¿ç”¨ç»Ÿè®¡"""
        return {
            'request_count': config['request_count'],
            'last_request': config['last_request']
        }
    
    def reset_stats():
        """é‡ç½®ç»Ÿè®¡"""
        config['request_count'] = 0
        config['last_request'] = None
    
    return {
        'request': make_request,
        'config': get_config,
        'update': update_config,
        'stats': get_stats,
        'reset': reset_stats
    }

# åˆ›å»ºAPIå®¢æˆ·ç«¯
api_client = create_api_client(
    base_url="https://api.example.com",
    default_headers={"Content-Type": "application/json"},
    timeout=60
)

print(f"   APIå®¢æˆ·ç«¯æµ‹è¯•:")
print(f"     åˆå§‹é…ç½®: {api_client['config']()}")

# å‘èµ·è¯·æ±‚
response1 = api_client['request']('/users', method='GET')
print(f"     GETè¯·æ±‚: {response1['url']}")

response2 = api_client['request']('/posts', method='POST', data={'title': 'Hello'}, headers={'Authorization': 'Bearer token'})
print(f"     POSTè¯·æ±‚: {response2['url']}")

print(f"     ä½¿ç”¨ç»Ÿè®¡: {api_client['stats']()}")

# æ›´æ–°é…ç½®
api_client['update'](timeout=120, default_headers={"User-Agent": "MyClient/1.0"})
print(f"     æ›´æ–°åé…ç½®: {api_client['config']()}")

# æ¡ˆä¾‹3ï¼šæ•°æ®å¤„ç†ç®¡é“å·¥å‚
print(f"\nğŸ“ 3. æ•°æ®å¤„ç†ç®¡é“å·¥å‚:")

def create_pipeline(initial_data=None):
    """åˆ›å»ºæ•°æ®å¤„ç†ç®¡é“å·¥å‚"""
    data = initial_data.copy() if initial_data else []
    operations = []  # æ“ä½œå†å²
    
    def add_operation(name, func, *args, **kwargs):
        """æ·»åŠ æ“ä½œåˆ°ç®¡é“"""
        operation = {
            'name': name,
            'func': func,
            'args': args,
            'kwargs': kwargs
        }
        operations.append(operation)
        print(f"     æ·»åŠ æ“ä½œ: {name}")
    
    def execute():
        """æ‰§è¡Œç®¡é“ä¸­çš„æ‰€æœ‰æ“ä½œ"""
        result = data.copy()
        
        for i, op in enumerate(operations, 1):
            try:
                print(f"     æ‰§è¡Œæ“ä½œ {i}: {op['name']}")
                result = op['func'](result, *op['args'], **op['kwargs'])
                print(f"       ç»“æœ: {len(result)} é¡¹")
            except Exception as e:
                print(f"       é”™è¯¯: {e}")
                break
        
        return result
    
    def get_operations():
        """è·å–æ“ä½œåˆ—è¡¨"""
        return [op['name'] for op in operations]
    
    def clear_operations():
        """æ¸…ç©ºæ“ä½œåˆ—è¡¨"""
        operations.clear()
    
    def set_data(new_data):
        """è®¾ç½®åˆå§‹æ•°æ®"""
        nonlocal data
        data = new_data.copy() if new_data else []
    
    return {
        'add': add_operation,
        'execute': execute,
        'operations': get_operations,
        'clear': clear_operations,
        'set_data': set_data
    }

# å®šä¹‰æ•°æ®å¤„ç†å‡½æ•°
def filter_positive(data):
    return [x for x in data if isinstance(x, (int, float)) and x > 0]

def remove_duplicates(data):
    seen = set()
    result = []
    for item in data:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

def sort_data(data, reverse=False):
    return sorted(data, reverse=reverse)

def transform_data(data, multiplier=1):
    return [x * multiplier for x in data if isinstance(x, (int, float))]

# åˆ›å»ºå¹¶æµ‹è¯•ç®¡é“
pipeline = create_pipeline([3, -1, 4, 2, -5, 4, 3, 1, 0])

print(f"   æ•°æ®å¤„ç†ç®¡é“æµ‹è¯•:")
print(f"     åˆå§‹æ•°æ®: {pipeline['set_data']([3, -1, 4, 2, -5, 4, 3, 1, 0])}")

# æ·»åŠ æ“ä½œ
pipeline['add']('è¿‡æ»¤æ­£æ•°', filter_positive)
pipeline['add']('å»é‡', remove_duplicates)
pipeline['add']('æ’åº', sort_data)
pipeline['add']('å˜æ¢æ•°æ®', transform_data, multiplier=2)

print(f"\n     æ“ä½œåˆ—è¡¨: {pipeline['operations']()}")

# æ‰§è¡Œç®¡é“
final_result = pipeline['execute']()
print(f"\n     æœ€ç»ˆç»“æœ: {final_result}")

# æ¡ˆä¾‹4ï¼šäº‹ä»¶ç³»ç»Ÿ
print(f"\nğŸ“ 4. äº‹ä»¶ç³»ç»Ÿ:")

def create_event_system():
    """åˆ›å»ºäº‹ä»¶ç³»ç»Ÿ"""
    listeners = {}  # äº‹ä»¶ç›‘å¬å™¨
    event_history = []  # äº‹ä»¶å†å²
    
    def on(event_name, callback):
        """æ³¨å†Œäº‹ä»¶ç›‘å¬å™¨"""
        if event_name not in listeners:
            listeners[event_name] = []
        listeners[event_name].append(callback)
        print(f"     æ³¨å†Œç›‘å¬å™¨: {event_name}")
    
    def emit(event_name, *args, **kwargs):
        """è§¦å‘äº‹ä»¶"""
        event_record = {
            'name': event_name,
            'args': args,
            'kwargs': kwargs,
            'timestamp': len(event_history)
        }
        event_history.append(event_record)
        
        print(f"     è§¦å‘äº‹ä»¶: {event_name}")
        
        if event_name in listeners:
            for callback in listeners[event_name]:
                try:
                    callback(*args, **kwargs)
                except Exception as e:
                    print(f"       ç›‘å¬å™¨é”™è¯¯: {e}")
    
    def off(event_name, callback=None):
        """ç§»é™¤äº‹ä»¶ç›‘å¬å™¨"""
        if event_name in listeners:
            if callback:
                if callback in listeners[event_name]:
                    listeners[event_name].remove(callback)
                    print(f"     ç§»é™¤ç‰¹å®šç›‘å¬å™¨: {event_name}")
            else:
                listeners[event_name].clear()
                print(f"     ç§»é™¤æ‰€æœ‰ç›‘å¬å™¨: {event_name}")
    
    def get_history():
        """è·å–äº‹ä»¶å†å²"""
        return event_history.copy()
    
    def clear_history():
        """æ¸…ç©ºäº‹ä»¶å†å²"""
        event_history.clear()
    
    return {
        'on': on,
        'emit': emit,
        'off': off,
        'history': get_history,
        'clear': clear_history
    }

# åˆ›å»ºäº‹ä»¶ç³»ç»Ÿ
event_system = create_event_system()

# å®šä¹‰äº‹ä»¶å¤„ç†å‡½æ•°
def on_user_login(username):
    print(f"       ç”¨æˆ·ç™»å½•: {username}")

def on_user_logout(username):
    print(f"       ç”¨æˆ·ç™»å‡º: {username}")

def on_data_update(data_type, count):
    print(f"       æ•°æ®æ›´æ–°: {data_type}, æ•°é‡: {count}")

print(f"   äº‹ä»¶ç³»ç»Ÿæµ‹è¯•:")

# æ³¨å†Œç›‘å¬å™¨
event_system['on']('user_login', on_user_login)
event_system['on']('user_logout', on_user_logout)
event_system['on']('data_update', on_data_update)

# è§¦å‘äº‹ä»¶
event_system['emit']('user_login', 'å¼ ä¸‰')
event_system['emit']('data_update', 'ç”¨æˆ·æ•°æ®', 100)
event_system['emit']('user_logout', 'å¼ ä¸‰')

# æŸ¥çœ‹äº‹ä»¶å†å²
print(f"\n     äº‹ä»¶å†å²: {event_system['history']()}")

print(f"\nâœ… å®é™…åº”ç”¨æ¡ˆä¾‹å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡è¿”å›å€¼å’Œä½œç”¨åŸŸå®é™…åº”ç”¨")
print(f"   âœ“ ç†è§£å‡½æ•°å¼ç¼–ç¨‹æ ¸å¿ƒæ¦‚å¿µ")
print(f"   âœ“ èƒ½è®¾è®¡ä¼˜é›…å‡½æ•°æ¥å£")
````
