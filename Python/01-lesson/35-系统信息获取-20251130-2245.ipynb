````python
print("âš™ï¸ osæ¨¡å—ç³»ç»Ÿçº§æ“ä½œ:")
print("=" * 50)

import os
import sys
import platform
import time
import tempfile
import json
from typing import Dict, List, Any, Optional, Union
from dataclasses import dataclass
from datetime import datetime
import pathlib

# 1. ç¯å¢ƒå˜é‡ç®¡ç†
print(f"ğŸ“ 1. ç¯å¢ƒå˜é‡ç®¡ç†:")

# 1.1 ç¯å¢ƒå˜é‡æ“ä½œ
print(f"\n   ğŸŒ 1.1 ç¯å¢ƒå˜é‡æ“ä½œ:")

@dataclass
class EnvironmentManager:
    """ç¯å¢ƒå˜é‡ç®¡ç†å™¨"""
    
    @staticmethod
    def get_all_env_vars() -> Dict[str, str]:
        """è·å–æ‰€æœ‰ç¯å¢ƒå˜é‡"""
        return dict(os.environ)
    
    @staticmethod
    def get_env_var(key: str, default: str = None) -> Optional[str]:
        """è·å–ç¯å¢ƒå˜é‡"""
        return os.environ.get(key, default)
    
    @staticmethod
    def set_env_var(key: str, value: str):
        """è®¾ç½®ç¯å¢ƒå˜é‡"""
        os.environ[key] = value
    
    @staticmethod
    def delete_env_var(key: str):
        """åˆ é™¤ç¯å¢ƒå˜é‡"""
        if key in os.environ:
            del os.environ[key]
    
    @staticmethod
    def list_langchain_env_vars() -> Dict[str, str]:
        """åˆ—å‡ºLangChainç›¸å…³ç¯å¢ƒå˜é‡"""
        langchain_vars = {}
        for key, value in os.environ.items():
            if any(keyword in key.upper() for keyword in ['LANGCHAIN', 'OPENAI', 'ANTHROPIC', 'HUGGINGFACE']):
                langchain_vars[key] = value
        return langchain_vars

def demonstrate_environment_management():
    """æ¼”ç¤ºç¯å¢ƒå˜é‡ç®¡ç†"""
    env_manager = EnvironmentManager()
    
    print(f"   ç¯å¢ƒå˜é‡ç®¡ç†æ¼”ç¤º:")
    
    # è·å–æ‰€æœ‰ç¯å¢ƒå˜é‡
    all_vars = env_manager.get_all_env_vars()
    print(f"\n   æ€»ç¯å¢ƒå˜é‡æ•°é‡: {len(all_vars)}")
    
    # æ˜¾ç¤ºé‡è¦ç¯å¢ƒå˜é‡
    important_vars = ['PATH', 'HOME', 'USER', 'PYTHONPATH', 'VIRTUAL_ENV']
    print(f"\n   é‡è¦ç¯å¢ƒå˜é‡:")
    
    for var in important_vars:
        value = env_manager.get_env_var(var)
        if value:
            if var == 'PATH':
                paths = value.split(os.pathsep)[:3]  # åªæ˜¾ç¤ºå‰3ä¸ªè·¯å¾„
                print(f"   {var}: {os.pathsep.join(paths)}... ({len(value.split(os.pathsep))}ä¸ªè·¯å¾„)")
            else:
                print(f"   {var}: {value}")
    
    # LangChainç›¸å…³å˜é‡
    langchain_vars = env_manager.list_langchain_env_vars()
    print(f"\n   LangChainç›¸å…³ç¯å¢ƒå˜é‡ ({len(langchain_vars)}ä¸ª):")
    for key, value in langchain_vars.items():
        # éšè—æ•æ„Ÿä¿¡æ¯
        if any(keyword in key.upper() for keyword in ['KEY', 'TOKEN', 'SECRET']):
            masked_value = value[:8] + '*' * (len(value) - 8) if len(value) > 8 else '*' * len(value)
            print(f"   {key}: {masked_value}")
        else:
            print(f"   {key}: {value}")
    
    # è®¾ç½®å’Œåˆ é™¤æµ‹è¯•å˜é‡
    print(f"\n   ç¯å¢ƒå˜é‡æ“ä½œæµ‹è¯•:")
    
    test_key = "PYTHON_TEST_VAR"
    test_value = "test_value_123"
    
    env_manager.set_env_var(test_key, test_value)
    retrieved_value = env_manager.get_env_var(test_key)
    print(f"   è®¾ç½® {test_key} = {test_value}")
    print(f"   è·å– {test_key} = {retrieved_value}")
    
    env_manager.delete_env_var(test_key)
    retrieved_after_delete = env_manager.get_env_var(test_key)
    print(f"   åˆ é™¤åè·å– {test_key} = {retrieved_after_delete}")

demonstrate_environment_management()

# 1.2 ç¯å¢ƒé…ç½®ç®¡ç†
print(f"\n   ğŸ”§ 1.2 ç¯å¢ƒé…ç½®ç®¡ç†:")

class EnvironmentConfig:
    """ç¯å¢ƒé…ç½®ç®¡ç†ç±»"""
    
    def __init__(self, config_file: str = None):
        self.config_file = config_file
        self.config = {}
        self.load_config()
    
    def load_config(self):
        """åŠ è½½é…ç½®"""
        if self.config_file and os.path.exists(self.config_file):
            try:
                with open(self.config_file, 'r', encoding='utf-8') as f:
                    self.config = json.load(f)
            except Exception as e:
                print(f"   é…ç½®æ–‡ä»¶åŠ è½½å¤±è´¥: {e}")
                self.config = {}
        else:
            # ä»ç¯å¢ƒå˜é‡åŠ è½½
            self.config = {
                'python_version': platform.python_version(),
                'platform': platform.system(),
                'user_home': os.path.expanduser('~'),
                'temp_dir': tempfile.gettempdir(),
                'current_dir': os.getcwd()
            }
    
    def get_config(self, key: str, default: Any = None) -> Any:
        """è·å–é…ç½®é¡¹"""
        return self.config.get(key, default)
    
    def set_config(self, key: str, value: Any):
        """è®¾ç½®é…ç½®é¡¹"""
        self.config[key] = value
    
    def save_config(self):
        """ä¿å­˜é…ç½®"""
        if self.config_file:
            try:
                with open(self.config_file, 'w', encoding='utf-8') as f:
                    json.dump(self.config, f, indent=2, ensure_ascii=False)
            except Exception as e:
                print(f"   é…ç½®æ–‡ä»¶ä¿å­˜å¤±è´¥: {e}")
    
    def detect_environment_type(self) -> str:
        """æ£€æµ‹ç¯å¢ƒç±»å‹"""
        # æ£€æµ‹è™šæ‹Ÿç¯å¢ƒ
        in_venv = hasattr(sys, 'real_prefix') or (hasattr(sys, 'base_prefix') and sys.base_prefix != sys.prefix)
        
        # æ£€æµ‹å®¹å™¨ç¯å¢ƒ
        in_container = os.path.exists('/.dockerenv') or os.path.exists('/proc/1/cgroup')
        
        # æ£€æµ‹CIç¯å¢ƒ
        ci_indicators = ['CI', 'GITHUB_ACTIONS', 'JENKINS_URL', 'TRAVIS']
        in_ci = any(os.environ.get(indicator) for indicator in ci_indicators)
        
        if in_ci:
            return 'CI/CD'
        elif in_container:
            return 'Container'
        elif in_venv:
            return 'Virtual Environment'
        else:
            return 'Native'

def demonstrate_environment_config():
    """æ¼”ç¤ºç¯å¢ƒé…ç½®ç®¡ç†"""
    # åˆ›å»ºä¸´æ—¶é…ç½®æ–‡ä»¶
    temp_config = tempfile.NamedTemporaryFile(mode='w', suffix='.json', delete=False)
    temp_config.close()
    
    try:
        config = EnvironmentConfig(temp_config.name)
        
        print(f"   ç¯å¢ƒé…ç½®ç®¡ç†æ¼”ç¤º:")
        
        # æ˜¾ç¤ºå½“å‰é…ç½®
        print(f"\n   å½“å‰ç¯å¢ƒé…ç½®:")
        for key, value in config.config.items():
            print(f"   {key}: {value}")
        
        # æ£€æµ‹ç¯å¢ƒç±»å‹
        env_type = config.detect_environment_type()
        print(f"\n   ç¯å¢ƒç±»å‹: {env_type}")
        
        # æ·»åŠ è‡ªå®šä¹‰é…ç½®
        config.set_config('langchain_version', '0.1.0')
        config.set_config('debug_mode', True)
        config.set_config('cache_dir', os.path.join(tempfile.gettempdir(), 'langchain_cache'))
        
        # ä¿å­˜é…ç½®
        config.save_config()
        print(f"\n   é…ç½®å·²ä¿å­˜åˆ°: {temp_config.name}")
        
        # é‡æ–°åŠ è½½éªŒè¯
        new_config = EnvironmentConfig(temp_config.name)
        print(f"\n   é‡æ–°åŠ è½½çš„é…ç½®:")
        for key, value in new_config.config.items():
            print(f"   {key}: {value}")
    
    finally:
        if os.path.exists(temp_config.name):
            os.unlink(temp_config.name)
            print(f"\n   ä¸´æ—¶é…ç½®æ–‡ä»¶å·²æ¸…ç†")

demonstrate_environment_config()

# 2. è¿›ç¨‹å’Œç³»ç»Ÿç®¡ç†
print(f"\nğŸ“ 2. è¿›ç¨‹å’Œç³»ç»Ÿç®¡ç†:")

# 2.1 è¿›ç¨‹ä¿¡æ¯è·å–
print(f"\n   ğŸ”„ 2.1 è¿›ç¨‹ä¿¡æ¯è·å–:")

@dataclass
class ProcessInfo:
    """è¿›ç¨‹ä¿¡æ¯"""
    pid: int
    ppid: int
    name: str
    current_dir: str
    user: str
    environment: Dict[str, str]

def get_process_info() -> ProcessInfo:
    """è·å–å½“å‰è¿›ç¨‹ä¿¡æ¯"""
    return ProcessInfo(
        pid=os.getpid(),
        ppid=os.getppid(),
        name=sys.argv[0] if sys.argv else 'python',
        current_dir=os.getcwd(),
        user=getpass.getuser(),
        environment=dict(os.environ)
    )

def demonstrate_process_info():
    """æ¼”ç¤ºè¿›ç¨‹ä¿¡æ¯è·å–"""
    process_info = get_process_info()
    
    print(f"   è¿›ç¨‹ä¿¡æ¯:")
    print(f"   è¿›ç¨‹ID: {process_info.pid}")
    print(f"   çˆ¶è¿›ç¨‹ID: {process_info.ppid}")
    print(f"   è¿›ç¨‹å: {process_info.name}")
    print(f"   å½“å‰ç›®å½•: {process_info.current_dir}")
    print(f"   ç”¨æˆ·: {process_info.user}")
    print(f"   ç¯å¢ƒå˜é‡æ•°é‡: {len(process_info.environment)}")
    
    # ç³»ç»Ÿè°ƒç”¨ç¤ºä¾‹
    print(f"\n   ç³»ç»Ÿè°ƒç”¨ç¤ºä¾‹:")
    
    # è·å–ç”¨æˆ·ä¿¡æ¯
    try:
        uid = os.getuid() if hasattr(os, 'getuid') else 'N/A (Windows)'
        gid = os.getgid() if hasattr(os, 'getgid') else 'N/A (Windows)'
        print(f"   ç”¨æˆ·ID: {uid}")
        print(f"   ç»„ID: {gid}")
    except AttributeError:
        print(f"   Windowsç³»ç»Ÿä¸æ”¯æŒuid/gid")
    
    # è·å–ç»ˆç«¯ä¿¡æ¯
    try:
        terminal = os.get_terminal_size()
        print(f"   ç»ˆç«¯å¤§å°: {terminal.columns} x {terminal.rows}")
    except OSError:
        print(f"   æ— æ³•è·å–ç»ˆç«¯å¤§å°")

demonstrate_process_info()

# 2.2 æ–‡ä»¶ç³»ç»Ÿæ“ä½œ
print(f"\n   ğŸ“ 2.2 æ–‡ä»¶ç³»ç»Ÿæ“ä½œ:")

class FileSystemManager:
    """æ–‡ä»¶ç³»ç»Ÿç®¡ç†å™¨"""
    
    @staticmethod
    def get_disk_usage(path: str = '/') -> Dict[str, int]:
        """è·å–ç£ç›˜ä½¿ç”¨æƒ…å†µ"""
        try:
            usage = shutil.disk_usage(path)
            return {
                'total': usage.total,
                'used': usage.used,
                'free': usage.free,
                'percent_used': (usage.used / usage.total) * 100 if usage.total > 0 else 0
            }
        except Exception as e:
            return {'error': str(e)}
    
    @staticmethod
    def format_bytes(bytes_value: int) -> str:
        """æ ¼å¼åŒ–å­—èŠ‚æ•°"""
        for unit in ['B', 'KB', 'MB', 'GB', 'TB']:
            if bytes_value < 1024.0:
                return f"{bytes_value:.2f} {unit}"
            bytes_value /= 1024.0
        return f"{bytes_value:.2f} PB"
    
    @staticmethod
    def analyze_directory(path: str, max_depth: int = 2) -> Dict[str, Any]:
        """åˆ†æç›®å½•ç»“æ„"""
        if not os.path.exists(path):
            return {'error': 'Path does not exist'}
        
        analysis = {
            'path': path,
            'file_count': 0,
            'dir_count': 0,
            'total_size': 0,
            'largest_files': [],
            'file_types': {}
        }
        
        try:
            for root, dirs, files in os.walk(path):
                # é™åˆ¶æ·±åº¦
                current_depth = root[len(path):].count(os.sep)
                if current_depth > max_depth:
                    dirs[:] = []  # ä¸ç»§ç»­æ·±å…¥
                    continue
                
                for file in files:
                    file_path = os.path.join(root, file)
                    try:
                        file_size = os.path.getsize(file_path)
                        analysis['file_count'] += 1
                        analysis['total_size'] += file_size
                        
                        # è®°å½•å¤§æ–‡ä»¶
                        analysis['largest_files'].append((file_path, file_size))
                        
                        # ç»Ÿè®¡æ–‡ä»¶ç±»å‹
                        ext = os.path.splitext(file)[1].lower()
                        if ext:
                            analysis['file_types'][ext] = analysis['file_types'].get(ext, 0) + 1
                        
                    except (OSError, PermissionError):
                        continue
                
                analysis['dir_count'] += len(dirs)
            
            # æ’åºæœ€å¤§çš„æ–‡ä»¶
            analysis['largest_files'].sort(key=lambda x: x[1], reverse=True)
            analysis['largest_files'] = analysis['largest_files'][:10]
            
        except (OSError, PermissionError) as e:
            analysis['error'] = str(e)
        
        return analysis

def demonstrate_filesystem_operations():
    """æ¼”ç¤ºæ–‡ä»¶ç³»ç»Ÿæ“ä½œ"""
    fs_manager = FileSystemManager()
    
    print(f"   æ–‡ä»¶ç³»ç»Ÿæ“ä½œæ¼”ç¤º:")
    
    # ç£ç›˜ä½¿ç”¨æƒ…å†µ
    current_path = os.getcwd()
    disk_usage = fs_manager.get_disk_usage(current_path)
    
    if 'error' not in disk_usage:
        print(f"\n   ç£ç›˜ä½¿ç”¨æƒ…å†µ ({current_path}):")
        print(f"   æ€»ç©ºé—´: {fs_manager.format_bytes(disk_usage['total'])}")
        print(f"   å·²ä½¿ç”¨: {fs_manager.format_bytes(disk_usage['used'])}")
        print(f"   å¯ç”¨ç©ºé—´: {fs_manager.format_bytes(disk_usage['free'])}")
        print(f"   ä½¿ç”¨ç‡: {disk_usage['percent_used']:.1f}%")
    else:
        print(f"   ç£ç›˜ä¿¡æ¯è·å–å¤±è´¥: {disk_usage['error']}")
    
    # åˆ†æå½“å‰ç›®å½•
    print(f"\n   å½“å‰ç›®å½•åˆ†æ:")
    dir_analysis = fs_manager.analyze_directory(current_path, max_depth=1)
    
    if 'error' not in dir_analysis:
        print(f"   æ–‡ä»¶æ•°é‡: {dir_analysis['file_count']}")
        print(f"   ç›®å½•æ•°é‡: {dir_analysis['dir_count']}")
        print(f"   æ€»å¤§å°: {fs_manager.format_bytes(dir_analysis['total_size'])}")
        
        if dir_analysis['largest_files']:
            print(f"\n   æœ€å¤§çš„æ–‡ä»¶:")
            for file_path, size in dir_analysis['largest_files'][:5]:
                file_name = os.path.basename(file_path)
                print(f"   {file_name}: {fs_manager.format_bytes(size)}")
        
        if dir_analysis['file_types']:
            print(f"\n   æ–‡ä»¶ç±»å‹ç»Ÿè®¡:")
            for ext, count in sorted(dir_analysis['file_types'].items(), key=lambda x: x[1], reverse=True)[:5]:
                print(f"   {ext}: {count}ä¸ª")
    else:
        print(f"   ç›®å½•åˆ†æå¤±è´¥: {dir_analysis['error']}")

demonstrate_filesystem_operations()

print(f"\nâœ… osæ¨¡å—ç³»ç»Ÿçº§æ“ä½œå®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡osæ¨¡å—çš„ç¯å¢ƒå˜é‡æ“ä½œ")
print(f"   âœ“ ç†è§£osæ¨¡å—çš„è¿›ç¨‹ç®¡ç†åŠŸèƒ½")
print(f"   âœ“ ç†Ÿç»ƒä½¿ç”¨osæ¨¡å—è¿›è¡Œæ–‡ä»¶ç³»ç»Ÿæ“ä½œ")
print(f"   âœ“ èƒ½å¤Ÿè¿›è¡Œè·¨å¹³å°ç³»ç»Ÿè°ƒç”¨")
````

````python
print("ğŸ” LangChainç¯å¢ƒæ£€æµ‹åº”ç”¨:")
print("=" * 50)

import platform
import sys
import os
import time
import tempfile
import json
import shutil
import gc
import importlib
import subprocess
from typing import Dict, List, Any, Optional, Union, Tuple
from dataclasses import dataclass, asdict
from datetime import datetime
from enum import Enum

# 1. ç¯å¢ƒæ£€æµ‹æ¡†æ¶
print(f"ğŸ“ 1. ç¯å¢ƒæ£€æµ‹æ¡†æ¶:")

# 1.1 æ£€æµ‹ç»“æœç±»å‹
print(f"\n   ğŸ·ï¸ 1.1 æ£€æµ‹ç»“æœç±»å‹:")

class CheckStatus(Enum):
    """æ£€æŸ¥çŠ¶æ€æšä¸¾"""
    PASS = "âœ… é€šè¿‡"
    WARNING = "âš ï¸ è­¦å‘Š"
    FAIL = "âŒ å¤±è´¥"
    SKIP = "â­ï¸ è·³è¿‡"

@dataclass
class CheckResult:
    """æ£€æŸ¥ç»“æœæ•°æ®ç±»"""
    name: str
    status: CheckStatus
    message: str
    details: Dict[str, Any] = None
    suggestion: str = None
    
    def __post_init__(self):
        if self.details is None:
            self.details = {}

@dataclass
class EnvironmentReport:
    """ç¯å¢ƒæŠ¥å‘Šæ•°æ®ç±»"""
    timestamp: datetime
    overall_status: CheckStatus
    checks: List[CheckResult]
    summary: Dict[str, int]
    recommendations: List[str]
    
    def to_dict(self) -> Dict[str, Any]:
        """è½¬æ¢ä¸ºå­—å…¸"""
        return {
            'timestamp': self.timestamp.isoformat(),
            'overall_status': self.overall_status.value,
            'checks': [asdict(check) for check in self.checks],
            'summary': self.summary,
            'recommendations': self.recommendations
        }

# 1.2 åŸºç¡€æ£€æŸ¥å™¨
print(f"\n   ğŸ”§ 1.2 åŸºç¡€æ£€æŸ¥å™¨:")

class BaseChecker:
    """åŸºç¡€æ£€æŸ¥å™¨ç±»"""
    
    def __init__(self, name: str):
        self.name = name
    
    def check(self) -> CheckResult:
        """æ‰§è¡Œæ£€æŸ¥ï¼Œå­ç±»éœ€è¦å®ç°"""
        raise NotImplementedError
    
    def _create_result(self, status: CheckStatus, message: str, 
                       details: Dict[str, Any] = None, suggestion: str = None) -> CheckResult:
        """åˆ›å»ºæ£€æŸ¥ç»“æœ"""
        return CheckResult(
            name=self.name,
            status=status,
            message=message,
            details=details,
            suggestion=suggestion
        )

# 2. å…·ä½“æ£€æŸ¥å™¨å®ç°
print(f"\nğŸ“ 2. å…·ä½“æ£€æŸ¥å™¨å®ç°:")

# 2.1 Pythonç¯å¢ƒæ£€æŸ¥å™¨
print(f"\n   ğŸ 2.1 Pythonç¯å¢ƒæ£€æŸ¥å™¨:")

class PythonEnvironmentChecker(BaseChecker):
    """Pythonç¯å¢ƒæ£€æŸ¥å™¨"""
    
    def __init__(self):
        super().__init__("Pythonç¯å¢ƒæ£€æŸ¥")
    
    def check(self) -> CheckResult:
        """æ£€æŸ¥Pythonç¯å¢ƒ"""
        details = {
            'python_version': platform.python_version(),
            'python_implementation': platform.python_implementation(),
            'python_executable': sys.executable,
            'platform': platform.system(),
            'architecture': platform.machine()
        }
        
        # æ£€æŸ¥Pythonç‰ˆæœ¬
        version_tuple = platform.python_version_tuple()
        major, minor = int(version_tuple[0]), int(version_tuple[1])
        
        if major < 3 or (major == 3 and minor < 8):
            return self._create_result(
                CheckStatus.FAIL,
                f"Pythonç‰ˆæœ¬è¿‡ä½: {platform.python_version()}",
                details,
                "è¯·å‡çº§åˆ°Python 3.8æˆ–æ›´é«˜ç‰ˆæœ¬"
            )
        elif major == 3 and minor < 10:
            return self._create_result(
                CheckStatus.WARNING,
                f"Pythonç‰ˆæœ¬å¯ç”¨ä½†å»ºè®®å‡çº§: {platform.python_version()}",
                details,
                "å»ºè®®å‡çº§åˆ°Python 3.10+ä»¥è·å¾—æ›´å¥½çš„æ€§èƒ½å’Œç‰¹æ€§"
            )
        else:
            return self._create_result(
                CheckStatus.PASS,
                f"Pythonç‰ˆæœ¬æ»¡è¶³è¦æ±‚: {platform.python_version()}",
                details
            )

# 2.2 å†…å­˜æ£€æŸ¥å™¨
print(f"\n   ğŸ’¾ 2.2 å†…å­˜æ£€æŸ¥å™¨:")

class MemoryChecker(BaseChecker):
    """å†…å­˜æ£€æŸ¥å™¨"""
    
    def __init__(self, min_memory_mb: int = 1024):
        super().__init__("å†…å­˜æ£€æŸ¥")
        self.min_memory_mb = min_memory_mb
    
    def check(self) -> CheckResult:
        """æ£€æŸ¥å†…å­˜çŠ¶å†µ"""
        # åƒåœ¾å›æ”¶
        gc.collect()
        
        # è·å–å†…å­˜ä¿¡æ¯ï¼ˆç®€åŒ–ç‰ˆæœ¬ï¼‰
        try:
            # å°è¯•è·å–ç³»ç»Ÿå†…å­˜ä¿¡æ¯
            if platform.system() == 'Windows':
                import ctypes
                kernel32 = ctypes.windll.kernel32
                c_ulonglong = ctypes.c_ulonglong
                
                class MEMORYSTATUSEX(ctypes.Structure):
                    _fields_ = [
                        ('dwLength', ctypes.c_ulong),
                        ('dwMemoryLoad', ctypes.c_ulong),
                        ('ullTotalPhys', c_ulonglong),
                        ('ullAvailPhys', c_ulonglong),
                        ('ullTotalPageFile', c_ulonglong),
                        ('ullAvailPageFile', c_ulonglong),
                        ('ullTotalVirtual', c_ulonglong),
                        ('ullAvailVirtual', c_ulonglong),
                        ('ullAvailExtendedVirtual', c_ulonglong)
                    ]
                
                memoryStatus = MEMORYSTATUSEX()
                memoryStatus.dwLength = ctypes.sizeof(MEMORYSTATUSEX)
                kernel32.GlobalMemoryStatusEx(ctypes.byref(memoryStatus))
                
                total_memory_mb = memoryStatus.ullTotalPhys // (1024 * 1024)
                available_memory_mb = memoryStatus.ullAvailPhys // (1024 * 1024)
                
            else:
                # Linux/Mac - ç®€åŒ–å¤„ç†
                try:
                    with open('/proc/meminfo', 'r') as f:
                        meminfo = f.read()
                    
                    # è§£æå†…å­˜ä¿¡æ¯
                    mem_total = 0
                    mem_available = 0
                    
                    for line in meminfo.split('\n'):
                        if 'MemTotal:' in line:
                            mem_total = int(line.split()[1]) // 1024  # KB to MB
                        elif 'MemAvailable:' in line:
                            mem_available = int(line.split()[1]) // 1024
                    
                    total_memory_mb = mem_total
                    available_memory_mb = mem_available
                    
                except:
                    # æ— æ³•è·å–ç³»ç»Ÿå†…å­˜ï¼Œä½¿ç”¨é»˜è®¤å€¼
                    total_memory_mb = 4096  # å‡è®¾4GB
                    available_memory_mb = 2048  # å‡è®¾2GBå¯ç”¨
            
            details = {
                'total_memory_mb': total_memory_mb,
                'available_memory_mb': available_memory_mb,
                'memory_usage_percent': ((total_memory_mb - available_memory_mb) / total_memory_mb) * 100
            }
            
            if available_memory_mb < self.min_memory_mb:
                return self._create_result(
                    CheckStatus.FAIL,
                    f"å¯ç”¨å†…å­˜ä¸è¶³: {available_memory_mb}MB < {self.min_memory_mb}MB",
                    details,
                    "è¯·é‡Šæ”¾å†…å­˜æˆ–å‡çº§åˆ°æ›´å¤§å†…å­˜çš„æœºå™¨"
                )
            elif available_memory_mb < self.min_memory_mb * 2:
                return self._create_result(
                    CheckStatus.WARNING,
                    f"å†…å­˜ç´§å¼ : {available_memory_mb}MB",
                    details,
                    "å»ºè®®é‡Šæ”¾ä¸€äº›å†…å­˜ä»¥è·å¾—æ›´å¥½æ€§èƒ½"
                )
            else:
                return self._create_result(
                    CheckStatus.PASS,
                    f"å†…å­˜å……è¶³: {available_memory_mb}MBå¯ç”¨",
                    details
                )
                
        except Exception as e:
            return self._create_result(
                CheckStatus.WARNING,
                f"æ— æ³•è·å–å†…å­˜ä¿¡æ¯: {str(e)}",
                {'error': str(e)},
                "è¯·æ‰‹åŠ¨æ£€æŸ¥ç³»ç»Ÿå†…å­˜çŠ¶å†µ"
            )

# 2.3 ç£ç›˜ç©ºé—´æ£€æŸ¥å™¨
print(f"\n   ğŸ’¿ 2.3 ç£ç›˜ç©ºé—´æ£€æŸ¥å™¨:")

class DiskSpaceChecker(BaseChecker):
    """ç£ç›˜ç©ºé—´æ£€æŸ¥å™¨"""
    
    def __init__(self, min_disk_gb: int = 5):
        super().__init__("ç£ç›˜ç©ºé—´æ£€æŸ¥")
        self.min_disk_gb = min_disk_gb
    
    def check(self) -> CheckResult:
        """æ£€æŸ¥ç£ç›˜ç©ºé—´"""
        try:
            current_path = os.getcwd()
            usage = shutil.disk_usage(current_path)
            
            total_gb = usage.total // (1024**3)
            free_gb = usage.free // (1024**3)
            used_gb = usage.used // (1024**3)
            usage_percent = (usage.used / usage.total) * 100
            
            details = {
                'total_gb': total_gb,
                'used_gb': used_gb,
                'free_gb': free_gb,
                'usage_percent': usage_percent,
                'check_path': current_path
            }
            
            if free_gb < self.min_disk_gb:
                return self._create_result(
                    CheckStatus.FAIL,
                    f"ç£ç›˜ç©ºé—´ä¸è¶³: {free_gb}GB < {self.min_disk_gb}GB",
                    details,
                    "è¯·æ¸…ç†ç£ç›˜ç©ºé—´æˆ–æ‰©å±•å­˜å‚¨å®¹é‡"
                )
            elif free_gb < self.min_disk_gb * 2:
                return self._create_result(
                    CheckStatus.WARNING,
                    f"ç£ç›˜ç©ºé—´ç´§å¼ : {free_gb}GBå¯ç”¨",
                    details,
                    "å»ºè®®æ¸…ç†ä¸€äº›æ–‡ä»¶ä»¥é‡Šæ”¾ç©ºé—´"
                )
            else:
                return self._create_result(
                    CheckStatus.PASS,
                    f"ç£ç›˜ç©ºé—´å……è¶³: {free_gb}GBå¯ç”¨",
                    details
                )
                
        except Exception as e:
            return self._create_result(
                CheckStatus.WARNING,
                f"æ— æ³•è·å–ç£ç›˜ä¿¡æ¯: {str(e)}",
                {'error': str(e)},
                "è¯·æ‰‹åŠ¨æ£€æŸ¥ç£ç›˜ç©ºé—´"
            )

# 2.4 ä¾èµ–åŒ…æ£€æŸ¥å™¨
print(f"\n   ğŸ“¦ 2.4 ä¾èµ–åŒ…æ£€æŸ¥å™¨:")

class DependencyChecker(BaseChecker):
    """ä¾èµ–åŒ…æ£€æŸ¥å™¨"""
    
    def __init__(self, required_packages: List[str] = None):
        super().__init__("ä¾èµ–åŒ…æ£€æŸ¥")
        self.required_packages = required_packages or [
            'requests', 'openai', 'tiktoken', 'numpy', 'pandas'
        ]
    
    def check(self) -> CheckResult:
        """æ£€æŸ¥ä¾èµ–åŒ…"""
        missing_packages = []
        installed_packages = []
        package_versions = {}
        
        for package in self.required_packages:
            try:
                module = importlib.import_module(package)
                version = getattr(module, '__version__', 'unknown')
                installed_packages.append(package)
                package_versions[package] = version
            except ImportError:
                missing_packages.append(package)
        
        details = {
            'required_count': len(self.required_packages),
            'installed_count': len(installed_packages),
            'missing_count': len(missing_packages),
            'installed_packages': installed_packages,
            'missing_packages': missing_packages,
            'package_versions': package_versions
        }
        
        if missing_packages:
            return self._create_result(
                CheckStatus.FAIL,
                f"ç¼ºå°‘{len(missing_packages)}ä¸ªä¾èµ–åŒ…: {', '.join(missing_packages)}",
                details,
                f"è¯·è¿è¡Œ: pip install {' '.join(missing_packages)}"
            )
        else:
            return self._create_result(
                CheckStatus.PASS,
                f"æ‰€æœ‰{len(installed_packages)}ä¸ªä¾èµ–åŒ…å·²å®‰è£…",
                details
            )

# 2.5 ç½‘ç»œè¿æ¥æ£€æŸ¥å™¨
print(f"\n   ğŸŒ 2.5 ç½‘ç»œè¿æ¥æ£€æŸ¥å™¨:")

class NetworkChecker(BaseChecker):
    """ç½‘ç»œè¿æ¥æ£€æŸ¥å™¨"""
    
    def __init__(self, test_urls: List[str] = None):
        super().__init__("ç½‘ç»œè¿æ¥æ£€æŸ¥")
        self.test_urls = test_urls or [
            'https://www.google.com',
            'https://api.openai.com',
            'https://pypi.org'
        ]
    
    def check(self) -> CheckResult:
        """æ£€æŸ¥ç½‘ç»œè¿æ¥"""
        import urllib.request
        import urllib.error
        
        reachable_urls = []
        unreachable_urls = []
        response_times = {}
        
        for url in self.test_urls:
            try:
                start_time = time.time()
                req = urllib.request.Request(url, method='HEAD')
                with urllib.request.urlopen(req, timeout=5) as response:
                    end_time = time.time()
                    response_time = (end_time - start_time) * 1000  # ms
                    
                    reachable_urls.append(url)
                    response_times[url] = response_time
                    
            except (urllib.error.URLError, urllib.error.HTTPError, Exception):
                unreachable_urls.append(url)
        
        details = {
            'total_urls': len(self.test_urls),
            'reachable_count': len(reachable_urls),
            'unreachable_count': len(unreachable_urls),
            'reachable_urls': reachable_urls,
            'unreachable_urls': unreachable_urls,
            'response_times': response_times
        }
        
        if not reachable_urls:
            return self._create_result(
                CheckStatus.FAIL,
                "ç½‘ç»œè¿æ¥å¤±è´¥ï¼Œæ— æ³•è®¿é—®ä»»ä½•æµ‹è¯•URL",
                details,
                "è¯·æ£€æŸ¥ç½‘ç»œè¿æ¥å’Œé˜²ç«å¢™è®¾ç½®"
            )
        elif len(unreachable_urls) > len(reachable_urls):
            return self._create_result(
                CheckStatus.WARNING,
                f"ç½‘ç»œè¿æ¥ä¸ç¨³å®š: {len(reachable_urls)}/{len(self.test_urls)} URLå¯è®¿é—®",
                details,
                "éƒ¨åˆ†ç½‘ç»œæœåŠ¡å¯èƒ½ä¸å¯ç”¨"
            )
        else:
            avg_response_time = sum(response_times.values()) / len(response_times) if response_times else 0
            return self._create_result(
                CheckStatus.PASS,
                f"ç½‘ç»œè¿æ¥æ­£å¸¸: {len(reachable_urls)}/{len(self.test_urls)} URLå¯è®¿é—®",
                {**details, 'avg_response_time_ms': avg_response_time}
            )

# 3. ç¯å¢ƒæ£€æµ‹ç®¡ç†å™¨
print(f"\nğŸ“ 3. ç¯å¢ƒæ£€æµ‹ç®¡ç†å™¨:")

class EnvironmentChecker:
    """ç¯å¢ƒæ£€æµ‹ç®¡ç†å™¨"""
    
    def __init__(self):
        self.checkers = [
            PythonEnvironmentChecker(),
            MemoryChecker(min_memory_mb=1024),
            DiskSpaceChecker(min_disk_gb=5),
            DependencyChecker(),
            NetworkChecker()
        ]
    
    def run_all_checks(self) -> EnvironmentReport:
        """è¿è¡Œæ‰€æœ‰æ£€æŸ¥"""
        print(f"   å¼€å§‹ç¯å¢ƒæ£€æµ‹...")
        
        results = []
        status_counts = {
            CheckStatus.PASS.value: 0,
            CheckStatus.WARNING.value: 0,
            CheckStatus.FAIL.value: 0,
            CheckStatus.SKIP.value: 0
        }
        
        for checker in self.checkers:
            print(f"   æ‰§è¡Œ: {checker.name}")
            result = checker.check()
            results.append(result)
            status_counts[result.status.value] += 1
            print(f"   ç»“æœ: {result.status.value} - {result.message}")
        
        # ç¡®å®šæ•´ä½“çŠ¶æ€
        if status_counts[CheckStatus.FAIL.value] > 0:
            overall_status = CheckStatus.FAIL
        elif status_counts[CheckStatus.WARNING.value] > 0:
            overall_status = CheckStatus.WARNING
        else:
            overall_status = CheckStatus.PASS
        
        # ç”Ÿæˆå»ºè®®
        recommendations = self._generate_recommendations(results)
        
        return EnvironmentReport(
            timestamp=datetime.now(),
            overall_status=overall_status,
            checks=results,
            summary=status_counts,
            recommendations=recommendations
        )
    
    def _generate_recommendations(self, results: List[CheckResult]) -> List[str]:
        """ç”Ÿæˆå»ºè®®"""
        recommendations = []
        
        for result in results:
            if result.suggestion:
                recommendations.append(f"{result.name}: {result.suggestion}")
        
        # é€šç”¨å»ºè®®
        if not recommendations:
            recommendations.append("ç¯å¢ƒé…ç½®è‰¯å¥½ï¼Œå¯ä»¥å¼€å§‹ä½¿ç”¨LangChain")
        
        return recommendations
    
    def save_report(self, report: EnvironmentReport, filename: str):
        """ä¿å­˜æŠ¥å‘Š"""
        try:
            with open(filename, 'w', encoding='utf-8') as f:
                json.dump(report.to_dict(), f, indent=2, ensure_ascii=False)
            print(f"   æŠ¥å‘Šå·²ä¿å­˜åˆ°: {filename}")
        except Exception as e:
            print(f"   æŠ¥å‘Šä¿å­˜å¤±è´¥: {e}")

def demonstrate_environment_checker():
    """æ¼”ç¤ºç¯å¢ƒæ£€æµ‹å™¨"""
    checker = EnvironmentChecker()
    
    print(f"   LangChainç¯å¢ƒæ£€æµ‹æ¼”ç¤º:")
    
    # è¿è¡Œæ£€æµ‹
    report = checker.run_all_checks()
    
    # æ˜¾ç¤ºæŠ¥å‘Š
    print(f"\n   ç¯å¢ƒæ£€æµ‹æŠ¥å‘Š:")
    print(f"   æ£€æµ‹æ—¶é—´: {report.timestamp}")
    print(f"   æ•´ä½“çŠ¶æ€: {report.overall_status.value}")
    
    print(f"\n   æ£€æµ‹ç»“æœç»Ÿè®¡:")
    for status, count in report.summary.items():
        print(f"   {status}: {count}é¡¹")
    
    print(f"\n   è¯¦ç»†ç»“æœ:")
    for check in report.checks:
        print(f"   {check.status.value} {check.name}: {check.message}")
        if check.suggestion:
            print(f"      å»ºè®®: {check.suggestion}")
    
    print(f"\n   æ”¹è¿›å»ºè®®:")
    for i, rec in enumerate(report.recommendations, 1):
        print(f"   {i}. {rec}")
    
    # ä¿å­˜æŠ¥å‘Š
    temp_report = tempfile.NamedTemporaryFile(mode='w', suffix='.json', delete=False)
    temp_report.close()
    
    try:
        checker.save_report(report, temp_report.name)
        
        # æ˜¾ç¤ºæŠ¥å‘Šæ–‡ä»¶å¤§å°
        file_size = os.path.getsize(temp_report.name)
        print(f"\n   æŠ¥å‘Šæ–‡ä»¶å¤§å°: {file_size} å­—èŠ‚")
    
    finally:
        if os.path.exists(temp_report.name):
            os.unlink(temp_report.name)

demonstrate_environment_checker()

print(f"\nâœ… LangChainç¯å¢ƒæ£€æµ‹åº”ç”¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡ç»¼åˆç¯å¢ƒæ£€æµ‹çš„æ–¹æ³•")
print(f"   âœ“ ç†è§£èµ„æºè¯„ä¼°å’Œé™åˆ¶æ£€æŸ¥")
print(f"   âœ“ ç†Ÿç»ƒä½¿ç”¨é…ç½®ç®¡ç†å’ŒéªŒè¯")
print(f"   âœ“ èƒ½å¤Ÿæ„å»ºç¯å¢ƒå¥åº·æ£€æŸ¥ç³»ç»Ÿ")
````
