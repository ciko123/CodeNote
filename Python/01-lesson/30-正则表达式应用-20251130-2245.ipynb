````python
print("ğŸ”§ é«˜çº§æ­£åˆ™è¡¨è¾¾å¼åº”ç”¨:")
print("=" * 50)

import re
import time
from typing import List, Tuple, Dict, Optional, Pattern, Union
from dataclasses import dataclass
from enum import Enum
import json

# 1. å¤æ‚æ¨¡å¼æ„å»º
print(f"ğŸ“ 1. å¤æ‚æ¨¡å¼æ„å»º:")

# 1.1 é‚®ç®±éªŒè¯æ¨¡å¼
print(f"\n   ğŸ“§ 1.1 é‚®ç®±éªŒè¯æ¨¡å¼:")

class EmailValidator:
    """é‚®ç®±éªŒè¯å™¨"""
    
    def __init__(self):
        # ç®€å•é‚®ç®±æ¨¡å¼
        self.simple_pattern = re.compile(
            r'^[\w.%+-]+@[\w.-]+\.[A-Za-z]{2,}$'
        )
        
        # ä¸¥æ ¼é‚®ç®±æ¨¡å¼
        self.strict_pattern = re.compile(r"""
            ^                           # å¼€å§‹
            [a-zA-Z0-9]                # ç¬¬ä¸€ä¸ªå­—ç¬¦
            [a-zA-Z0-9._%+-]{0,63}     # ç”¨æˆ·åéƒ¨åˆ†
            @                           # @ç¬¦å·
            [a-zA-Z0-9]                # åŸŸåç¬¬ä¸€ä¸ªå­—ç¬¦
            [a-zA-Z0-9.-]{0,63}        # åŸŸåéƒ¨åˆ†
            \.[a-zA-Z]{2,}             # é¡¶çº§åŸŸå
            $                           # ç»“æŸ
        """, re.VERBOSE)
        
        # ä¼ä¸šé‚®ç®±æ¨¡å¼
        self.enterprise_pattern = re.compile(r"""
            ^                           # å¼€å§‹
            [a-zA-Z0-9._%+-]+           # ç”¨æˆ·å
            @                           # @ç¬¦å·
            (?:                         # åŸŸåç»„
                [a-zA-Z0-9-]+           # å­åŸŸå
                \.[a-zA-Z0-9-]+        # åŸŸå
            )+                          # ä¸€ä¸ªæˆ–å¤šä¸ªå­åŸŸå
            \.[a-zA-Z]{2,}             # é¡¶çº§åŸŸå
            $                           # ç»“æŸ
        """, re.VERBOSE)
    
    def validate_simple(self, email: str) -> bool:
        """ç®€å•éªŒè¯"""
        return bool(self.simple_pattern.match(email))
    
    def validate_strict(self, email: str) -> bool:
        """ä¸¥æ ¼éªŒè¯"""
        return bool(self.strict_pattern.match(email))
    
    def validate_enterprise(self, email: str) -> bool:
        """ä¼ä¸šé‚®ç®±éªŒè¯"""
        return bool(self.enterprise_pattern.match(email))
    
    def extract_domain(self, email: str) -> Optional[str]:
        """æå–åŸŸå"""
        match = re.search(r'@([\w.-]+\.[A-Za-z]{2,})', email)
        return match.group(1) if match else None

# æµ‹è¯•é‚®ç®±éªŒè¯
validator = EmailValidator()
test_emails = [
    "user@example.com",
    "admin@test.org",
    "support@company.net",
    "invalid.email",
    "@missing-user.com",
    "user@.com",
    "very.long.username@sub.domain.company.com",
    "user+tag@example.org"
]

print(f"   é‚®ç®±éªŒè¯æµ‹è¯•:")
for email in test_emails:
    simple = validator.validate_simple(email)
    strict = validator.validate_strict(email)
    enterprise = validator.validate_enterprise(email)
    domain = validator.extract_domain(email)
    print(f"     {email:30} | ç®€å•: {'âœ“' if simple else 'âœ—'} | ä¸¥æ ¼: {'âœ“' if strict else 'âœ—'} | ä¼ä¸š: {'âœ“' if enterprise else 'âœ—'} | åŸŸå: {domain}")

# 1.2 URLè§£ææ¨¡å¼
print(f"\n   ğŸŒ 1.2 URLè§£ææ¨¡å¼:")

class URLParser:
    """URLè§£æå™¨"""
    
    def __init__(self):
        # URLæ¨¡å¼
        self.url_pattern = re.compile(r"""
            ^(?:(?P<scheme>https?)://)?      # åè®®
            (?:(?P<username>[^:@]+)(?::(?P<password>[^@]*))?@)?  # ç”¨æˆ·åå¯†ç 
            (?P<host>[^/:?#]+)               # ä¸»æœº
            (?::(?P<port>\d+))?             # ç«¯å£
            (?P<path>/[^?#]*)?               # è·¯å¾„
            (?:\?(?P<query>[^#]*))?         # æŸ¥è¯¢å‚æ•°
            (?:\#(?P<fragment>.*))?         # ç‰‡æ®µ
            $                               # ç»“æŸ
        """, re.VERBOSE)
        
        # IPåœ°å€æ¨¡å¼
        self.ip_pattern = re.compile(
            r'^(?:(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.){3}'
            r'(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)$'
        )
    
    def parse_url(self, url: str) -> Optional[Dict[str, str]]:
        """è§£æURL"""
        match = self.url_pattern.match(url)
        if match:
            return match.groupdict()
        return None
    
    def is_ip_address(self, host: str) -> bool:
        """åˆ¤æ–­æ˜¯å¦ä¸ºIPåœ°å€"""
        return bool(self.ip_pattern.match(host))
    
    def extract_emails_from_url(self, url: str) -> List[str]:
        """ä»URLä¸­æå–é‚®ç®±"""
        email_pattern = r'[\w.%+-]+@[\w.-]+\.[A-Za-z]{2,}'
        return re.findall(email_pattern, url)

# æµ‹è¯•URLè§£æ
parser = URLParser()
test_urls = [
    "https://www.example.com:8080/path/to/page?param1=value1&param2=value2#section",
    "http://user:password@example.org/login",
    "ftp://192.168.1.1/files",
    "https://api.test.com/v1/users",
    "www.simple.com",
    "mailto:user@example.com"
]

print(f"   URLè§£ææµ‹è¯•:")
for url in test_urls:
    parsed = parser.parse_url(url)
    if parsed:
        host = parsed.get('host', '')
        is_ip = parser.is_ip_address(host)
        print(f"\n     URL: {url}")
        print(f"       åè®®: {parsed.get('scheme', 'None')}")
        print(f"       ä¸»æœº: {host} ({'IPåœ°å€' if is_ip else 'åŸŸå'})")
        print(f"       ç«¯å£: {parsed.get('port', 'None')}")
        print(f"       è·¯å¾„: {parsed.get('path', 'None')}")
        print(f"       æŸ¥è¯¢: {parsed.get('query', 'None')}")
    else:
        print(f"\n     URL: {url} - è§£æå¤±è´¥")

# 1.3 æ—¥å¿—è§£ææ¨¡å¼
print(f"\n   ğŸ“‹ 1.3 æ—¥å¿—è§£ææ¨¡å¼:")

class LogParser:
    """æ—¥å¿—è§£æå™¨"""
    
    def __init__(self):
        # Apacheæ—¥å¿—æ¨¡å¼
        self.apache_pattern = re.compile(r"""
            ^(?P<ip>\S+)\s+                    # IPåœ°å€
            (?P<identity>\S+)\s+              # èº«ä»½
            (?P<user>\S+)\s+                  # ç”¨æˆ·
            \[(?P<time>[^]]+)\]\s+            # æ—¶é—´
            "(?P<method>\S+)\s+               # æ–¹æ³•
            (?P<path>\S+)\s+                  # è·¯å¾„
            (?P<protocol>\S+)"\s+             # åè®®
            (?P<status>\d+)\s+                # çŠ¶æ€ç 
            (?P<size>\S+)\s+                  # å¤§å°
            "(?P<referer>[^"]*)"\s+          # å¼•ç”¨é¡µ
            "(?P<user_agent>[^"]*)"          # ç”¨æˆ·ä»£ç†
            $                                 # ç»“æŸ
        """, re.VERBOSE)
        
        # é”™è¯¯æ—¥å¿—æ¨¡å¼
        self.error_pattern = re.compile(r"""
            ^(?P<timestamp>\d{4}-\d{2}-\d{2}\s+\d{2}:\d{2}:\d{2})  # æ—¶é—´æˆ³
            \s+(?P<level>\w+)\s+              # æ—¥å¿—çº§åˆ«
            \[(?P<module>[^]]+)\]\s+          # æ¨¡å—
            (?P<message>.*)                    # æ¶ˆæ¯
            $                                   # ç»“æŸ
        """, re.VERBOSE)
    
    def parse_apache_log(self, log_line: str) -> Optional[Dict[str, str]]:
        """è§£æApacheæ—¥å¿—"""
        match = self.apache_pattern.match(log_line)
        return match.groupdict() if match else None
    
    def parse_error_log(self, log_line: str) -> Optional[Dict[str, str]]:
        """è§£æé”™è¯¯æ—¥å¿—"""
        match = self.error_pattern.match(log_line)
        return match.groupdict() if match else None
    
    def extract_ip_addresses(self, text: str) -> List[str]:
        """æå–IPåœ°å€"""
        ip_pattern = r'\b(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\b'
        return re.findall(ip_pattern, text)
    
    def filter_by_status(self, logs: List[str], status_codes: List[int]) -> List[Dict[str, str]]:
        """æŒ‰çŠ¶æ€ç è¿‡æ»¤æ—¥å¿—"""
        filtered = []
        for log in logs:
            parsed = self.parse_apache_log(log)
            if parsed and int(parsed.get('status', 0)) in status_codes:
                filtered.append(parsed)
        return filtered

# æµ‹è¯•æ—¥å¿—è§£æ
log_parser = LogParser()
apache_logs = [
    '192.168.1.100 - - [25/Dec/2024:10:30:45 +0000] "GET /index.html HTTP/1.1" 200 1024 "http://example.com" "Mozilla/5.0"',
    '10.0.0.1 - user [25/Dec/2024:10:31:00 +0000] "POST /login HTTP/1.1" 302 0 "http://example.com/login" "Mozilla/5.0"',
    '172.16.0.50 - - [25/Dec/2024:10:31:15 +0000] "GET /api/users HTTP/1.1" 404 512 "http://example.com" "curl/7.68.0"',
    '203.0.113.1 - admin [25/Dec/2024:10:31:30 +0000] "DELETE /user/123 HTTP/1.1" 500 256 "http://example.com/admin" "Python-requests/2.25.1"'
]

error_logs = [
    '2024-12-25 10:30:45 ERROR [database] Connection failed to localhost',
    '2024-12-25 10:31:00 WARNING [auth] Invalid login attempt for user admin',
    '2024-12-25 10:31:15 ERROR [api] Request timeout after 30 seconds',
    '2024-12-25 10:31:30 INFO [cache] Cache cleared successfully'
]

print(f"   Apacheæ—¥å¿—è§£æ:")
for log in apache_logs:
    parsed = log_parser.parse_apache_log(log)
    if parsed:
        print(f"\n     IP: {parsed['ip']} | æ–¹æ³•: {parsed['method']} | è·¯å¾„: {parsed['path']} | çŠ¶æ€: {parsed['status']}")

print(f"\n   é”™è¯¯æ—¥å¿—è§£æ:")
for log in error_logs:
    parsed = log_parser.parse_error_log(log)
    if parsed:
        print(f"     æ—¶é—´: {parsed['timestamp']} | çº§åˆ«: {parsed['level']} | æ¨¡å—: {parsed['module']} | æ¶ˆæ¯: {parsed['message']}")

# æå–IPåœ°å€
all_logs_text = '\n'.join(apache_logs)
ips = log_parser.extract_ip_addresses(all_logs_text)
print(f"\n   æå–çš„IPåœ°å€: {ips}")

# æŒ‰çŠ¶æ€ç è¿‡æ»¤
error_status_logs = log_parser.filter_by_status(apache_logs, [404, 500])
print(f"\n   é”™è¯¯çŠ¶æ€æ—¥å¿— (404, 500):")
for log in error_status_logs:
    print(f"     {log['ip']} -> {log['path']} ({log['status']})")

# 2. æ€§èƒ½ä¼˜åŒ–
print(f"\nğŸ“ 2. æ€§èƒ½ä¼˜åŒ–:")

# 2.1 ç¼–è¯‘æ­£åˆ™è¡¨è¾¾å¼
print(f"\n   âš¡ 2.1 ç¼–è¯‘æ­£åˆ™è¡¨è¾¾å¼:")

def test_regex_performance():
    """æµ‹è¯•æ­£åˆ™è¡¨è¾¾å¼æ€§èƒ½"""
    # ç”Ÿæˆæµ‹è¯•æ•°æ®
    test_data = ['user{}@example.com'.format(i) for i in range(1000)]
    test_data.extend(['invalid{}email'.format(i) for i in range(1000)])
    
    # æœªç¼–è¯‘çš„æ­£åˆ™è¡¨è¾¾å¼
    pattern_str = r'^[\w.%+-]+@[\w.-]+\.[A-Za-z]{2,}$'
    
    # æµ‹è¯•æœªç¼–è¯‘ç‰ˆæœ¬
    start_time = time.perf_counter()
    for email in test_data:
        re.match(pattern_str, email)
    uncompiled_time = time.perf_counter() - start_time
    
    # æµ‹è¯•ç¼–è¯‘ç‰ˆæœ¬
    compiled_pattern = re.compile(pattern_str)
    start_time = time.perf_counter()
    for email in test_data:
        compiled_pattern.match(email)
    compiled_time = time.perf_counter() - start_time
    
    return uncompiled_time, compiled_time

uncompiled_time, compiled_time = test_regex_performance()
print(f"   æ€§èƒ½æµ‹è¯• (2000æ¬¡åŒ¹é…):")
print(f"   æœªç¼–è¯‘ç‰ˆæœ¬: {uncompiled_time:.6f} ç§’")
print(f"   ç¼–è¯‘ç‰ˆæœ¬: {compiled_time:.6f} ç§’")
print(f"   æ€§èƒ½æå‡: {uncompiled_time/compiled_time:.2f}x")

# 2.2 ä¼˜åŒ–æ¨¡å¼è®¾è®¡
print(f"\n   ğŸ”§ 2.2 ä¼˜åŒ–æ¨¡å¼è®¾è®¡:")

class OptimizedPatterns:
    """ä¼˜åŒ–çš„æ­£åˆ™è¡¨è¾¾å¼æ¨¡å¼"""
    
    def __init__(self):
        # è´ªå©ªæ¨¡å¼ vs éè´ªå©ªæ¨¡å¼
        self.greedy_pattern = re.compile(r'<.*>')
        self.non_greedy_pattern = re.compile(r'<.*?>')
        
        # å›æº¯ä¼˜åŒ–
        self.backtrack_pattern = re.compile(r'^(a+)+b$')  # å®¹æ˜“å›æº¯
        self.optimized_pattern = re.compile(r'^a+b$')    # ä¼˜åŒ–å
        
        # å­—ç¬¦ç±»ä¼˜åŒ–
        self.inefficient_pattern = re.compile(r'(?:a|b|c|d|e|f|g|h|i|j)')
        self.efficient_pattern = re.compile(r'[a-j]')
    
    def test_greedy_vs_non_greedy(self):
        """æµ‹è¯•è´ªå©ªvséè´ªå©ª"""
        text = '<div>content</div><span>more</span>'
        
        greedy_match = self.greedy_pattern.search(text)
        non_greedy_match = self.non_greedy_pattern.search(text)
        
        print(f"   æ–‡æœ¬: {text}")
        print(f"   è´ªå©ªåŒ¹é…: {greedy_match.group() if greedy_match else 'None'}")
        print(f"   éè´ªå©ªåŒ¹é…: {non_greedy_match.group() if non_greedy_match else 'None'}")
    
    def test_backtrack_optimization(self):
        """æµ‹è¯•å›æº¯ä¼˜åŒ–"""
        test_strings = ['ab', 'aaaaab', 'aaaaaaaaaaaaaaaaab']
        
        print(f"\n   å›æº¯ä¼˜åŒ–æµ‹è¯•:")
        for test_str in test_strings:
            # æµ‹è¯•å®¹æ˜“å›æº¯çš„æ¨¡å¼
            start_time = time.perf_counter()
            backtrack_match = self.backtrack_pattern.match(test_str)
            backtrack_time = time.perf_counter() - start_time
            
            # æµ‹è¯•ä¼˜åŒ–åçš„æ¨¡å¼
            start_time = time.perf_counter()
            optimized_match = self.optimized_pattern.match(test_str)
            optimized_time = time.perf_counter() - start_time
            
            print(f"     å­—ç¬¦ä¸²: {test_str:20} | å›æº¯: {backtrack_time:.6f}s | ä¼˜åŒ–: {optimized_time:.6f}s")

# æµ‹è¯•ä¼˜åŒ–æ¨¡å¼
optimized = OptimizedPatterns()
optimized.test_greedy_vs_non_greedy()
optimized.test_backtrack_optimization()

# 3. å®é™…åº”ç”¨åœºæ™¯
print(f"\nğŸ“ 3. å®é™…åº”ç”¨åœºæ™¯:")

# 3.1 æ•°æ®æ¸…æ´—
print(f"\n   ğŸ§¹ 3.1 æ•°æ®æ¸…æ´—:")

class DataCleaner:
    """æ•°æ®æ¸…æ´—å™¨"""
    
    def __init__(self):
        # æ¸…ç†æ¨¡å¼
        self.whitespace_pattern = re.compile(r'\s+')
        self.phone_pattern = re.compile(r'[\s\-\(\)]')
        self.currency_pattern = re.compile(r'[^\d.]')
        self.html_pattern = re.compile(r'<[^>]+>')
        self.special_char_pattern = re.compile(r'[^a-zA-Z0-9\s]')
    
    def normalize_whitespace(self, text: str) -> str:
        """è§„èŒƒåŒ–ç©ºç™½å­—ç¬¦"""
        return self.whitespace_pattern.sub(' ', text).strip()
    
    def clean_phone_number(self, phone: str) -> str:
        """æ¸…ç†ç”µè¯å·ç """
        return self.phone_pattern.sub('', phone)
    
    def extract_amount(self, currency_text: str) -> float:
        """æå–é‡‘é¢"""
        clean_text = self.currency_pattern.sub('', currency_text)
        try:
            return float(clean_text)
        except ValueError:
            return 0.0
    
    def remove_html_tags(self, html_text: str) -> str:
        """ç§»é™¤HTMLæ ‡ç­¾"""
        return self.html_pattern.sub('', html_text)
    
    def extract_keywords(self, text: str) -> List[str]:
        """æå–å…³é”®è¯"""
        # ç§»é™¤ç‰¹æ®Šå­—ç¬¦
        clean_text = self.special_char_pattern.sub(' ', text)
        # åˆ†å‰²å•è¯
        words = clean_text.split()
        # è¿‡æ»¤çŸ­è¯å’Œåœç”¨è¯
        stop_words = {'the', 'a', 'an', 'and', 'or', 'but', 'in', 'on', 'at', 'to', 'for', 'of', 'with', 'by'}
        keywords = [word.lower() for word in words if len(word) > 2 and word.lower() not in stop_words]
        return list(set(keywords))

# æµ‹è¯•æ•°æ®æ¸…æ´—
cleaner = DataCleaner()

# æµ‹è¯•æ•°æ®
messy_text = "  Hello    World!   This   is   a   test.  "
phone_numbers = ["(123) 456-7890", "123-456-7890", "123 456 7890", "1234567890"]
currency_texts = ["$1,234.56", "â‚¬1.234,56", "Â¥1,234", "1234.56 USD"]
html_text = "<p>Hello <b>World</b>!</p><div>This is <i>HTML</i> content.</div>"
article_text = "The quick brown fox jumps over the lazy dog. Python programming is powerful and easy to learn."

print(f"   æ•°æ®æ¸…æ´—æµ‹è¯•:")
print(f"\n   ç©ºç™½å­—ç¬¦è§„èŒƒåŒ–:")
print(f"     åŸæ–‡: {repr(messy_text)}")
print(f"     æ¸…ç†å: {repr(cleaner.normalize_whitespace(messy_text))}")

print(f"\n   ç”µè¯å·ç æ¸…ç†:")
for phone in phone_numbers:
    cleaned = cleaner.clean_phone_number(phone)
    print(f"     {phone:15} -> {cleaned}")

print(f"\n   é‡‘é¢æå–:")
for currency in currency_texts:
    amount = cleaner.extract_amount(currency)
    print(f"     {currency:15} -> {amount}")

print(f"\n   HTMLæ ‡ç­¾ç§»é™¤:")
print(f"     åŸæ–‡: {html_text}")
print(f"     æ¸…ç†å: {cleaner.remove_html_tags(html_text)}")

print(f"\n   å…³é”®è¯æå–:")
keywords = cleaner.extract_keywords(article_text)
print(f"     åŸæ–‡: {article_text}")
print(f"     å…³é”®è¯: {keywords}")

# 3.2 æ–‡æœ¬åˆ†æ
print(f"\n   ğŸ“Š 3.2 æ–‡æœ¬åˆ†æ:")

class TextAnalyzer:
    """æ–‡æœ¬åˆ†æå™¨"""
    
    def __init__(self):
        # åˆ†ææ¨¡å¼
        self.word_pattern = re.compile(r'\b[a-zA-Z]+\b')
        self.sentence_pattern = re.compile(r'[.!?]+')
        self.email_pattern = re.compile(r'\b[\w.%+-]+@[\w.-]+\.[A-Za-z]{2,}\b')
        self.url_pattern = re.compile(r'https?://[^\s]+')
        self.number_pattern = re.compile(r'\b\d+(?:\.\d+)?\b')
    
    def word_count(self, text: str) -> int:
        """ç»Ÿè®¡å•è¯æ•°é‡"""
        words = self.word_pattern.findall(text)
        return len(words)
    
    def sentence_count(self, text: str) -> int:
        """ç»Ÿè®¡å¥å­æ•°é‡"""
        sentences = self.sentence_pattern.split(text)
        return len([s for s in sentences if s.strip()])
    
    def extract_emails(self, text: str) -> List[str]:
        """æå–é‚®ç®±"""
        return self.email_pattern.findall(text)
    
    def extract_urls(self, text: str) -> List[str]:
        """æå–URL"""
        return self.url_pattern.findall(text)
    
    def extract_numbers(self, text: str) -> List[float]:
        """æå–æ•°å­—"""
        numbers = self.number_pattern.findall(text)
        return [float(n) for n in numbers]
    
    def analyze_text(self, text: str) -> Dict[str, Union[int, List]]:
        """ç»¼åˆåˆ†ææ–‡æœ¬"""
        return {
            'word_count': self.word_count(text),
            'sentence_count': self.sentence_count(text),
            'emails': self.extract_emails(text),
            'urls': self.extract_urls(text),
            'numbers': self.extract_numbers(text)
        }

# æµ‹è¯•æ–‡æœ¬åˆ†æ
analyzer = TextAnalyzer()
sample_text = """
Hello World! This is a sample text for analysis. 
You can contact us at support@example.com or admin@test.org.
Visit our website at https://www.example.com for more information.
The price is $99.99 and the discount is 15.5%. Phone: 123-456-7890.
This text contains 3 sentences and multiple data points.
"""

analysis = analyzer.analyze_text(sample_text)
print(f"   æ–‡æœ¬åˆ†æç»“æœ:")
for key, value in analysis.items():
    if isinstance(value, list):
        print(f"     {key}: {value}")
    else:
        print(f"     {key}: {value}")

print(f"\nâœ… é«˜çº§æ­£åˆ™è¡¨è¾¾å¼åº”ç”¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å¤æ‚æ­£åˆ™è¡¨è¾¾å¼æ¨¡å¼çš„æ„å»ºæŠ€å·§")
print(f"   âœ“ ç†è§£æ­£åˆ™è¡¨è¾¾å¼çš„æ€§èƒ½ä¼˜åŒ–æ–¹æ³•")
print(f"   âœ“ èƒ½å¤„ç†å®é™…åº”ç”¨ä¸­çš„æ–‡æœ¬å¤„ç†åœºæ™¯")
print(f"   âœ“ æŒæ¡æ­£åˆ™è¡¨è¾¾å¼çš„è°ƒè¯•å’Œæµ‹è¯•æŠ€å·§")
````
