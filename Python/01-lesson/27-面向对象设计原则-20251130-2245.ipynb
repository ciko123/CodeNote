````python
print("ğŸ—ï¸ SOLIDè®¾è®¡åŸåˆ™:")
print("=" * 50)

# 1. å•ä¸€èŒè´£åŸåˆ™ (Single Responsibility Principle)
print(f"ğŸ“ 1. å•ä¸€èŒè´£åŸåˆ™ (SRP):")
print(f"   ä¸€ä¸ªç±»åº”è¯¥åªæœ‰ä¸€ä¸ªå¼•èµ·å˜åŒ–çš„åŸå› ")

# è¿åSRPçš„ä¾‹å­
print(f"\n   âŒ è¿åSRPçš„ä¾‹å­:")

class BadUserManager:
    """è¿åå•ä¸€èŒè´£åŸåˆ™çš„ç”¨æˆ·ç®¡ç†å™¨"""
    
    def __init__(self):
        self.users = []
    
    def add_user(self, name, email):
        """æ·»åŠ ç”¨æˆ·"""
        user = {'name': name, 'email': email}
        self.users.append(user)
        print(f"   æ·»åŠ ç”¨æˆ·: {name}")
    
    def save_to_database(self):
        """ä¿å­˜åˆ°æ•°æ®åº“"""
        print(f"   ä¿å­˜ç”¨æˆ·åˆ°æ•°æ®åº“")
    
    def send_email(self, user, message):
        """å‘é€é‚®ä»¶"""
        print(f"   å‘é€é‚®ä»¶ç»™ {user['name']}: {message}")
    
    def generate_report(self):
        """ç”ŸæˆæŠ¥å‘Š"""
        print(f"   ç”Ÿæˆç”¨æˆ·æŠ¥å‘Š")

# éµå¾ªSRPçš„ä¾‹å­
print(f"\n   âœ… éµå¾ªSRPçš„ä¾‹å­:")

class User:
    """ç”¨æˆ·å®ä½“ç±»"""
    
    def __init__(self, name, email):
        self.name = name
        self.email = email
    
    def __str__(self):
        return f"User({self.name}, {self.email})"

class UserRepository:
    """ç”¨æˆ·æ•°æ®ä»“åº“"""
    
    def __init__(self):
        self.users = []
    
    def add(self, user):
        self.users.append(user)
        print(f"   æ·»åŠ ç”¨æˆ·åˆ°ä»“åº“: {user}")
    
    def save_to_database(self):
        print(f"   ä¿å­˜ç”¨æˆ·åˆ°æ•°æ®åº“")
    
    def get_all(self):
        return self.users.copy()

class EmailService:
    """é‚®ä»¶æœåŠ¡"""
    
    def send_email(self, user, message):
        print(f"   å‘é€é‚®ä»¶ç»™ {user.name}: {message}")
        return f"é‚®ä»¶å·²å‘é€ç»™ {user.name}"

class UserReportService:
    """ç”¨æˆ·æŠ¥å‘ŠæœåŠ¡"""
    
    def generate_report(self, users):
        print(f"   ç”Ÿæˆç”¨æˆ·æŠ¥å‘Šï¼Œå…± {len(users)} ä¸ªç”¨æˆ·")
        return f"ç”¨æˆ·æŠ¥å‘Š: {len(users)} ä¸ªç”¨æˆ·"

class UserManager:
    """ç”¨æˆ·ç®¡ç†å™¨ - åè°ƒå„ä¸ªæœåŠ¡"""
    
    def __init__(self, repository, email_service, report_service):
        self.repository = repository
        self.email_service = email_service
        self.report_service = report_service
    
    def add_user(self, name, email):
        user = User(name, email)
        self.repository.add(user)
        return user
    
    def notify_user(self, user, message):
        return self.email_service.send_email(user, message)
    
    def generate_user_report(self):
        users = self.repository.get_all()
        return self.report_service.generate_report(users)

# æµ‹è¯•SRP
print(f"\n   æµ‹è¯•å•ä¸€èŒè´£åŸåˆ™:")

# åˆ›å»ºæœåŠ¡å®ä¾‹
repository = UserRepository()
email_service = EmailService()
report_service = UserReportService()
user_manager = UserManager(repository, email_service, report_service)

# æ·»åŠ ç”¨æˆ·
user1 = user_manager.add_user("å¼ ä¸‰", "zhangsan@example.com")
user2 = user_manager.add_user("æå››", "lisi@example.com")

# å‘é€é€šçŸ¥
user_manager.notify_user(user1, "æ¬¢è¿æ³¨å†Œï¼")

# ç”ŸæˆæŠ¥å‘Š
report = user_manager.generate_user_report()
print(f"       {report}")

# 2. å¼€é—­åŸåˆ™ (Open/Closed Principle)
print(f"\nğŸ“ 2. å¼€é—­åŸåˆ™ (OCP):")
print(f"   è½¯ä»¶å®ä½“åº”è¯¥å¯¹æ‰©å±•å¼€æ”¾ï¼Œå¯¹ä¿®æ”¹å…³é—­")

# è¿åOCPçš„ä¾‹å­
print(f"\n   âŒ è¿åOCPçš„ä¾‹å­:")

class BadShapeCalculator:
    """è¿åå¼€é—­åŸåˆ™çš„å½¢çŠ¶è®¡ç®—å™¨"""
    
    def calculate_area(self, shape_type, *args):
        if shape_type == "circle":
            radius = args[0]
            return 3.14 * radius * radius
        elif shape_type == "rectangle":
            width, height = args
            return width * height
        elif shape_type == "triangle":
            base, height = args
            return 0.5 * base * height
        else:
            raise ValueError(f"ä¸æ”¯æŒçš„å½¢çŠ¶: {shape_type}")

# éµå¾ªOCPçš„ä¾‹å­
print(f"\n   âœ… éµå¾ªOCPçš„ä¾‹å­:")

from abc import ABC, abstractmethod

class Shape(ABC):
    """å½¢çŠ¶æŠ½è±¡åŸºç±»"""
    
    @abstractmethod
    def calculate_area(self):
        pass
    
    @abstractmethod
    def calculate_perimeter(self):
        pass

class Circle(Shape):
    """åœ†å½¢"""
    
    def __init__(self, radius):
        self.radius = radius
    
    def calculate_area(self):
        return 3.14 * self.radius * self.radius
    
    def calculate_perimeter(self):
        return 2 * 3.14 * self.radius

class Rectangle(Shape):
    """çŸ©å½¢"""
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def calculate_area(self):
        return self.width * self.height
    
    def calculate_perimeter(self):
        return 2 * (self.width + self.height)

class Triangle(Shape):
    """ä¸‰è§’å½¢"""
    
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def calculate_area(self):
        return 0.5 * self.base * self.height
    
    def calculate_perimeter(self):
        # ç®€åŒ–è®¡ç®—ï¼Œå‡è®¾ç­‰è¾¹ä¸‰è§’å½¢
        side = (self.base ** 2 + self.height ** 2) ** 0.5
        return self.base + 2 * side

class ShapeCalculator:
    """å½¢çŠ¶è®¡ç®—å™¨ - å¯¹æ‰©å±•å¼€æ”¾ï¼Œå¯¹ä¿®æ”¹å…³é—­"""
    
    def calculate_total_area(self, shapes):
        total_area = 0
        for shape in shapes:
            total_area += shape.calculate_area()
        return total_area
    
    def calculate_total_perimeter(self, shapes):
        total_perimeter = 0
        for shape in shapes:
            total_perimeter += shape.calculate_perimeter()
        return total_perimeter
    
    def get_shape_info(self, shapes):
        info = []
        for i, shape in enumerate(shapes):
            shape_type = shape.__class__.__name__
            area = shape.calculate_area()
            perimeter = shape.calculate_perimeter()
            info.append({
                'index': i,
                'type': shape_type,
                'area': area,
                'perimeter': perimeter
            })
        return info

# æµ‹è¯•OCP
print(f"\n   æµ‹è¯•å¼€é—­åŸåˆ™:")

# åˆ›å»ºå½¢çŠ¶å¯¹è±¡
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4)
]

# åˆ›å»ºè®¡ç®—å™¨
calculator = ShapeCalculator()

# è®¡ç®—æ€»é¢ç§¯å’Œå‘¨é•¿
total_area = calculator.calculate_total_area(shapes)
total_perimeter = calculator.calculate_total_perimeter(shapes)

print(f"\n     å½¢çŠ¶ä¿¡æ¯:")
shape_info = calculator.get_shape_info(shapes)
for info in shape_info:
    print(f"       {info['type']}: é¢ç§¯={info['area']:.2f}, å‘¨é•¿={info['perimeter']:.2f}")

print(f"\n     æ€»é¢ç§¯: {total_area:.2f}")
print(f"     æ€»å‘¨é•¿: {total_perimeter:.2f}")

# 3. é‡Œæ°æ›¿æ¢åŸåˆ™ (Liskov Substitution Principle)
print(f"\nğŸ“ 3. é‡Œæ°æ›¿æ¢åŸåˆ™ (LSP):")
print(f"   å­ç±»å¯¹è±¡åº”è¯¥èƒ½å¤Ÿæ›¿æ¢çˆ¶ç±»å¯¹è±¡")

# è¿åLSPçš„ä¾‹å­
print(f"\n   âŒ è¿åLSPçš„ä¾‹å­:")

class BadBird:
    """é¸Ÿç±»åŸºç±»"""
    
    def fly(self):
        return "é¸Ÿåœ¨é£ç¿”"

class BadPenguin(BadBird):
    """ä¼é¹…ç±» - è¿åé‡Œæ°æ›¿æ¢åŸåˆ™"""
    
    def fly(self):
        raise RuntimeError("ä¼é¹…ä¸ä¼šé£")

# éµå¾ªLSPçš„ä¾‹å­
print(f"\n   âœ… éµå¾ªLSPçš„ä¾‹å­:")

class Bird:
    """é¸Ÿç±»åŸºç±»"""
    
    def make_sound(self):
        return "é¸Ÿå«"
    
    def move(self):
        return "é¸Ÿåœ¨ç§»åŠ¨"

class FlyingBird(Bird):
    """ä¼šé£çš„é¸Ÿç±»"""
    
    def fly(self):
        return "é¸Ÿåœ¨é£ç¿”"
    
    def move(self):
        return self.fly()

class WalkingBird(Bird):
    """ä¼šèµ°çš„é¸Ÿç±»"""
    
    def walk(self):
        return "é¸Ÿåœ¨è¡Œèµ°"
    
    def move(self):
        return self.walk()

class Sparrow(FlyingBird):
    """éº»é›€"""
    
    def make_sound(self):
        return "éº»é›€å«"

class Penguin(WalkingBird):
    """ä¼é¹…"""
    
    def make_sound(self):
        return "ä¼é¹…å«"

class BirdHandler:
    """é¸Ÿç±»å¤„ç†å™¨"""
    
    def handle_bird(self, bird):
        sound = bird.make_sound()
        movement = bird.move()
        return f"{sound}, {movement}"
    
    def process_birds(self, birds):
        results = []
        for bird in birds:
            result = self.handle_bird(bird)
            results.append({
                'type': bird.__class__.__name__,
                'result': result
            })
        return results

# æµ‹è¯•LSP
print(f"\n   æµ‹è¯•é‡Œæ°æ›¿æ¢åŸåˆ™:")

# åˆ›å»ºé¸Ÿç±»å¯¹è±¡
birds = [
    Sparrow(),
    Penguin(),
    FlyingBird(),
    WalkingBird()
]

# åˆ›å»ºå¤„ç†å™¨
handler = BirdHandler()

# å¤„ç†æ‰€æœ‰é¸Ÿç±»
results = handler.process_birds(birds)

print(f"\n     å¤„ç†ç»“æœ:")
for result in results:
    print(f"       {result['type']}: {result['result']}")

print(f"\nâœ… SOLIDåŸåˆ™ç¬¬ä¸€éƒ¨åˆ†å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å•ä¸€èŒè´£åŸåˆ™(SRP)çš„åº”ç”¨")
print(f"   âœ“ ç†è§£å¼€é—­åŸåˆ™(OCP)çš„è®¾è®¡æ€æƒ³")
print(f"   âœ“ èƒ½è¿ç”¨é‡Œæ°æ›¿æ¢åŸåˆ™(LSP)è®¾è®¡ç»§æ‰¿å±‚æ¬¡")
````

### SOLIDåŸåˆ™ç»­ä¸å…¶ä»–è®¾è®¡åŸåˆ™ [â­â­è¿›é˜¶]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šç»§ç»­å­¦ä¹ SOLIDåŸåˆ™çš„å‰©ä½™éƒ¨åˆ†ï¼Œä»¥åŠå…¶ä»–é‡è¦çš„é¢å‘å¯¹è±¡è®¾è®¡åŸåˆ™ï¼Œå¦‚DRYã€KISSã€YAGNIç­‰ã€‚

````python
print("ğŸ”§ SOLIDåŸåˆ™ç»­ä¸å…¶ä»–è®¾è®¡åŸåˆ™:")
print("=" * 50)

# 4. æ¥å£éš”ç¦»åŸåˆ™ (Interface Segregation Principle)
print(f"ğŸ“ 4. æ¥å£éš”ç¦»åŸåˆ™ (ISP):")
print(f"   å®¢æˆ·ç«¯ä¸åº”è¯¥ä¾èµ–å®ƒä¸éœ€è¦çš„æ¥å£")

# è¿åISPçš„ä¾‹å­
print(f"\n   âŒ è¿åISPçš„ä¾‹å­:")

class BadWorker:
    """è¿åæ¥å£éš”ç¦»åŸåˆ™çš„å·¥äººæ¥å£"""
    
    def work(self):
        pass
    
    def eat(self):
        pass
    
    def sleep(self):
        pass

class BadRobot(BadWorker):
    """æœºå™¨äºº - ä¸éœ€è¦eatå’Œsleepæ–¹æ³•"""
    
    def work(self):
        return "æœºå™¨äººåœ¨å·¥ä½œ"
    
    def eat(self):
        raise NotImplementedError("æœºå™¨äººä¸éœ€è¦åƒé¥­")
    
    def sleep(self):
        raise NotImplementedError("æœºå™¨äººä¸éœ€è¦ç¡è§‰")

# éµå¾ªISPçš„ä¾‹å­
print(f"\n   âœ… éµå¾ªISPçš„ä¾‹å­:")

from abc import ABC, abstractmethod

class Workable(ABC):
    """å¯å·¥ä½œæ¥å£"""
    
    @abstractmethod
    def work(self):
        pass

class Eatable(ABC):
    """å¯è¿›é£Ÿæ¥å£"""
    
    @abstractmethod
    def eat(self):
        pass

class Sleepable(ABC):
    """å¯ç¡çœ æ¥å£"""
    
    @abstractmethod
    def sleep(self):
        pass

class Human(Workable, Eatable, Sleepable):
    """äººç±» - å®ç°æ‰€æœ‰æ¥å£"""
    
    def work(self):
        return "äººç±»åœ¨å·¥ä½œ"
    
    def eat(self):
        return "äººç±»åœ¨åƒé¥­"
    
    def sleep(self):
        return "äººç±»åœ¨ç¡è§‰"

class Robot(Workable):
    """æœºå™¨äºº - åªå®ç°å·¥ä½œæ¥å£"""
    
    def work(self):
        return "æœºå™¨äººåœ¨å·¥ä½œ"

class Animal(Eatable, Sleepable):
    """åŠ¨ç‰© - åªå®ç°è¿›é£Ÿå’Œç¡çœ æ¥å£"""
    
    def eat(self):
        return "åŠ¨ç‰©åœ¨è¿›é£Ÿ"
    
    def sleep(self):
        return "åŠ¨ç‰©åœ¨ç¡è§‰"

class WorkerManager:
    """å·¥äººç®¡ç†å™¨"""
    
    def manage_work(self, worker: Workable):
        return worker.work()
    
    def manage_meal(self, eater: Eatable):
        return eater.eat()
    
    def manage_rest(self, sleeper: Sleepable):
        return sleeper.sleep()

# æµ‹è¯•ISP
print(f"\n   æµ‹è¯•æ¥å£éš”ç¦»åŸåˆ™:")

# åˆ›å»ºå¯¹è±¡
human = Human()
robot = Robot()
animal = Animal()

# åˆ›å»ºç®¡ç†å™¨
manager = WorkerManager()

# ç®¡ç†å·¥ä½œ
print(f"\n     å·¥ä½œç®¡ç†:")
print(f"       äººç±»: {manager.manage_work(human)}")
print(f"       æœºå™¨äºº: {manager.manage_work(robot)}")

# ç®¡ç†è¿›é£Ÿ
print(f"\n     è¿›é£Ÿç®¡ç†:")
print(f"       äººç±»: {manager.manage_meal(human)}")
print(f"       åŠ¨ç‰©: {manager.manage_meal(animal)}")

# ç®¡ç†ä¼‘æ¯
print(f"\n     ä¼‘æ¯ç®¡ç†:")
print(f"       äººç±»: {manager.manage_rest(human)}")
print(f"       åŠ¨ç‰©: {manager.manage_rest(animal)}")

# 5. ä¾èµ–å€’ç½®åŸåˆ™ (Dependency Inversion Principle)
print(f"\nğŸ“ 5. ä¾èµ–å€’ç½®åŸåˆ™ (DIP):")
print(f"   é«˜å±‚æ¨¡å—ä¸åº”è¯¥ä¾èµ–ä½å±‚æ¨¡å—ï¼Œä¸¤è€…éƒ½åº”è¯¥ä¾èµ–æŠ½è±¡")

# è¿åDIPçš„ä¾‹å­
print(f"\n   âŒ è¿åDIPçš„ä¾‹å­:")

class BadLightBulb:
    """ç¯æ³¡ç±»"""
    
    def turn_on(self):
        return "ç¯æ³¡äº®äº†"
    
    def turn_off(self):
        return "ç¯æ³¡ç­äº†"

class BadSwitch:
    """å¼€å…³ç±» - ç›´æ¥ä¾èµ–å…·ä½“ç¯æ³¡"""
    
    def __init__(self):
        self.bulb = BadLightBulb()
        self.is_on = False
    
    def toggle(self):
        if self.is_on:
            result = self.bulb.turn_off()
        else:
            result = self.bulb.turn_on()
        self.is_on = not self.is_on
        return result

# éµå¾ªDIPçš„ä¾‹å­
print(f"\n   âœ… éµå¾ªDIPçš„ä¾‹å­:")

from abc import ABC, abstractmethod

class Switchable(ABC):
    """å¯å¼€å…³è®¾å¤‡æ¥å£"""
    
    @abstractmethod
    def turn_on(self):
        pass
    
    @abstractmethod
    def turn_off(self):
        pass

class LightBulb(Switchable):
    """ç¯æ³¡ç±»"""
    
    def turn_on(self):
        return "ç¯æ³¡äº®äº†"
    
    def turn_off(self):
        return "ç¯æ³¡ç­äº†"

class Fan(Switchable):
    """é£æ‰‡ç±»"""
    
    def turn_on(self):
        return "é£æ‰‡è½¬äº†"
    
    def turn_off(self):
        return "é£æ‰‡åœäº†"

class Computer(Switchable):
    """ç”µè„‘ç±»"""
    
    def turn_on(self):
        return "ç”µè„‘å¯åŠ¨äº†"
    
    def turn_off(self):
        return "ç”µè„‘å…³é—­äº†"

class ElectricSwitch:
    """ç”µç¯å¼€å…³ - ä¾èµ–æŠ½è±¡æ¥å£"""
    
    def __init__(self, device: Switchable):
        self.device = device
        self.is_on = False
    
    def toggle(self):
        if self.is_on:
            result = self.device.turn_off()
        else:
            result = self.device.turn_on()
        self.is_on = not self.is_on
        return result
    
    def turn_on(self):
        if not self.is_on:
            result = self.device.turn_on()
            self.is_on = True
            return result
        return "è®¾å¤‡å·²ç»å¼€å¯"
    
    def turn_off(self):
        if self.is_on:
            result = self.device.turn_off()
            self.is_on = False
            return result
        return "è®¾å¤‡å·²ç»å…³é—­"

class SmartHome:
    """æ™ºèƒ½å®¶å±…ç³»ç»Ÿ"""
    
    def __init__(self):
        self.switches = []
    
    def add_switch(self, switch):
        self.switches.append(switch)
        print(f"   æ·»åŠ å¼€å…³: {switch.device.__class__.__name__}")
    
    def toggle_all(self):
        results = []
        for switch in self.switches:
            result = switch.toggle()
            results.append({
                'device': switch.device.__class__.__name__,
                'result': result
            })
        return results
    
    def turn_on_all(self):
        results = []
        for switch in self.switches:
            result = switch.turn_on()
            results.append({
                'device': switch.device.__class__.__name__,
                'result': result
            })
        return results

# æµ‹è¯•DIP
print(f"\n   æµ‹è¯•ä¾èµ–å€’ç½®åŸåˆ™:")

# åˆ›å»ºè®¾å¤‡
bulb = LightBulb()
fan = Fan()
computer = Computer()

# åˆ›å»ºå¼€å…³
bulb_switch = ElectricSwitch(bulb)
fan_switch = ElectricSwitch(fan)
computer_switch = ElectricSwitch(computer)

# åˆ›å»ºæ™ºèƒ½å®¶å±…
smart_home = SmartHome()
smart_home.add_switch(bulb_switch)
smart_home.add_switch(fan_switch)
smart_home.add_switch(computer_switch)

# æµ‹è¯•å¼€å…³
print(f"\n     æµ‹è¯•å•ä¸ªå¼€å…³:")
print(f"       ç¯æ³¡å¼€å…³: {bulb_switch.toggle()}")
print(f"       é£æ‰‡å¼€å…³: {fan_switch.toggle()}")
print(f"       ç”µè„‘å¼€å…³: {computer_switch.toggle()}")

# æµ‹è¯•æ™ºèƒ½å®¶å±…
print(f"\n     æµ‹è¯•æ™ºèƒ½å®¶å±…ç³»ç»Ÿ:")
results = smart_home.turn_on_all()
for result in results:
    print(f"       {result['device']}: {result['result']}")

# 6. å…¶ä»–é‡è¦è®¾è®¡åŸåˆ™
print(f"\nğŸ“ 6. å…¶ä»–é‡è¦è®¾è®¡åŸåˆ™:")

# DRYåŸåˆ™ (Don't Repeat Yourself)
print(f"\n   ğŸ”„ DRYåŸåˆ™ (Don't Repeat Yourself):")
print(f"      ä¸è¦é‡å¤è‡ªå·±")

class DataProcessor:
    """æ•°æ®å¤„ç†å™¨ - éµå¾ªDRYåŸåˆ™"""
    
    def __init__(self):
        self.processed_count = 0
    
    def _validate_data(self, data, required_fields=None):
        """é€šç”¨æ•°æ®éªŒè¯æ–¹æ³•"""
        if not isinstance(data, dict):
            raise ValueError("æ•°æ®å¿…é¡»æ˜¯å­—å…¸ç±»å‹")
        
        if required_fields:
            missing_fields = [field for field in required_fields if field not in data]
            if missing_fields:
                raise ValueError(f"ç¼ºå°‘å¿…éœ€å­—æ®µ: {missing_fields}")
        
        return True
    
    def _log_operation(self, operation, data_id):
        """é€šç”¨æ—¥å¿—è®°å½•æ–¹æ³•"""
        import datetime
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"   [{timestamp}] {operation}: æ•°æ®ID {data_id}")
        self.processed_count += 1
    
    def process_user_data(self, data):
        """å¤„ç†ç”¨æˆ·æ•°æ®"""
        self._validate_data(data, ['name', 'email'])
        self._log_operation("å¤„ç†ç”¨æˆ·æ•°æ®", data.get('id', 'unknown'))
        # å¤„ç†é€»è¾‘
        return f"ç”¨æˆ· {data['name']} å¤„ç†å®Œæˆ"
    
    def process_product_data(self, data):
        """å¤„ç†äº§å“æ•°æ®"""
        self._validate_data(data, ['title', 'price'])
        self._log_operation("å¤„ç†äº§å“æ•°æ®", data.get('id', 'unknown'))
        # å¤„ç†é€»è¾‘
        return f"äº§å“ {data['title']} å¤„ç†å®Œæˆ"

# KISSåŸåˆ™ (Keep It Simple, Stupid)
print(f"\n   ğŸ¯ KISSåŸåˆ™ (Keep It Simple, Stupid):")
print(f"      ä¿æŒç®€å•")

class SimpleCalculator:
    """ç®€å•è®¡ç®—å™¨ - éµå¾ªKISSåŸåˆ™"""
    
    def add(self, a, b):
        return a + b
    
    def subtract(self, a, b):
        return a - b
    
    def multiply(self, a, b):
        return a * b
    
    def divide(self, a, b):
        if b == 0:
            raise ValueError("é™¤æ•°ä¸èƒ½ä¸ºé›¶")
        return a / b

# YAGNIåŸåˆ™ (You Aren't Gonna Need It)
print(f"\n   âš¡ YAGNIåŸåˆ™ (You Aren't Gonna Need It):")
print(f"      ä½ ä¸ä¼šéœ€è¦å®ƒ")

class UserAuth:
    """ç”¨æˆ·è®¤è¯ - éµå¾ªYAGNIåŸåˆ™"""
    
    def __init__(self):
        self.users = {
            'admin': 'password123',
            'user1': 'pass456'
        }
    
    def authenticate(self, username, password):
        """è®¤è¯ç”¨æˆ· - åªå®ç°å½“å‰éœ€è¦çš„åŠŸèƒ½"""
        if username in self.users and self.users[username] == password:
            return True
        return False
    
    # ä¸é¢„å…ˆå®ç°ä¸éœ€è¦çš„åŠŸèƒ½
    # def reset_password(self, username):
    #     pass  # å½“å‰ä¸éœ€è¦ï¼Œä¸å®ç°
    # def change_password(self, username, old_password, new_password):
    #     pass  # å½“å‰ä¸éœ€è¦ï¼Œä¸å®ç°

# æµ‹è¯•å…¶ä»–åŸåˆ™
print(f"\n   æµ‹è¯•å…¶ä»–è®¾è®¡åŸåˆ™:")

# æµ‹è¯•DRYåŸåˆ™
print(f"\n     æµ‹è¯•DRYåŸåˆ™:")
processor = DataProcessor()

user_data = {'id': 1, 'name': 'å¼ ä¸‰', 'email': 'zhangsan@example.com'}
product_data = {'id': 101, 'title': 'æ‰‹æœº', 'price': 2999}

result1 = processor.process_user_data(user_data)
result2 = processor.process_product_data(product_data)

print(f"       {result1}")
print(f"       {result2}")
print(f"       æ€»å¤„ç†æ•°é‡: {processor.processed_count}")

# æµ‹è¯•KISSåŸåˆ™
print(f"\n     æµ‹è¯•KISSåŸåˆ™:")
calculator = SimpleCalculator()

print(f"       10 + 5 = {calculator.add(10, 5)}")
print(f"       10 - 5 = {calculator.subtract(10, 5)}")
print(f"       10 * 5 = {calculator.multiply(10, 5)}")
print(f"       10 / 5 = {calculator.divide(10, 5)}")

# æµ‹è¯•YAGNIåŸåˆ™
print(f"\n     æµ‹è¯•YAGNIåŸåˆ™:")
auth = UserAuth()

print(f"       adminè®¤è¯: {auth.authenticate('admin', 'password123')}")
print(f"       user1è®¤è¯: {auth.authenticate('user1', 'wrongpass')}")

print(f"\nâœ… SOLIDåŸåˆ™ç»­ä¸å…¶ä»–è®¾è®¡åŸåˆ™å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡æ¥å£éš”ç¦»åŸåˆ™(ISP)çš„åº”ç”¨")
print(f"   âœ“ ç†è§£ä¾èµ–å€’ç½®åŸåˆ™(DIP)çš„è®¾è®¡æ€æƒ³")
print(f"   âœ“ èƒ½è¿ç”¨DRYã€KISSã€YAGNIç­‰åŸåˆ™ä¼˜åŒ–ä»£ç ")
````

````python
print("ğŸ—ï¸ è®¾è®¡åŸåˆ™ç»¼åˆåº”ç”¨:")
print("=" * 50)

# ç»¼åˆæ¡ˆä¾‹ï¼šç”µå•†è®¢å•å¤„ç†ç³»ç»Ÿ
print(f"ğŸ“ ç»¼åˆæ¡ˆä¾‹ï¼šç”µå•†è®¢å•å¤„ç†ç³»ç»Ÿ")
print(f"   åº”ç”¨å¤šç§è®¾è®¡åŸåˆ™æ„å»ºå¯ç»´æŠ¤çš„ç³»ç»Ÿ")

# 1. å®šä¹‰æŠ½è±¡æ¥å£ (DIP, ISP)
from abc import ABC, abstractmethod
from typing import List, Dict, Any

class OrderItem(ABC):
    """è®¢å•é¡¹æ¥å£"""
    
    @abstractmethod
    def calculate_price(self) -> float:
        pass
    
    @abstractmethod
    def get_description(self) -> str:
        pass

class PaymentProcessor(ABC):
    """æ”¯ä»˜å¤„ç†å™¨æ¥å£"""
    
    @abstractmethod
    def process_payment(self, amount: float) -> bool:
        pass

class NotificationService(ABC):
    """é€šçŸ¥æœåŠ¡æ¥å£"""
    
    @abstractmethod
    def send_notification(self, message: str, recipient: str) -> bool:
        pass

class OrderRepository(ABC):
    """è®¢å•ä»“åº“æ¥å£"""
    
    @abstractmethod
    def save_order(self, order: 'Order') -> bool:
        pass
    
    @abstractmethod
    def get_order(self, order_id: str) -> 'Order':
        pass

# 2. å®ç°å…·ä½“ç±» (SRP, OCP)
class ProductItem(OrderItem):
    """äº§å“è®¢å•é¡¹ - å•ä¸€èŒè´£"""
    
    def __init__(self, product_id: str, name: str, price: float, quantity: int):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def calculate_price(self) -> float:
        return self.price * self.quantity
    
    def get_description(self) -> str:
        return f"{self.name} x{self.quantity} (Â¥{self.price})"

class ServiceItem(OrderItem):
    """æœåŠ¡è®¢å•é¡¹ - æ‰©å±•æ–°çš„è®¢å•é¡¹ç±»å‹"""
    
    def __init__(self, service_id: str, name: str, price: float, duration: int):
        self.service_id = service_id
        self.name = name
        self.price = price
        self.duration = duration
    
    def calculate_price(self) -> float:
        return self.price * self.duration
    
    def get_description(self) -> str:
        return f"{self.name} ({self.duration}å°æ—¶) (Â¥{self.price}/å°æ—¶)"

# 3. æ”¯ä»˜å¤„ç†å™¨å®ç° (OCP, LSP)
class CreditCardProcessor(PaymentProcessor):
    """ä¿¡ç”¨å¡æ”¯ä»˜å¤„ç†å™¨"""
    
    def process_payment(self, amount: float) -> bool:
        print(f"   å¤„ç†ä¿¡ç”¨å¡æ”¯ä»˜: Â¥{amount}")
        # æ¨¡æ‹Ÿæ”¯ä»˜å¤„ç†
        return True

class AlipayProcessor(PaymentProcessor):
    """æ”¯ä»˜å®æ”¯ä»˜å¤„ç†å™¨"""
    
    def process_payment(self, amount: float) -> bool:
        print(f"   å¤„ç†æ”¯ä»˜å®æ”¯ä»˜: Â¥{amount}")
        # æ¨¡æ‹Ÿæ”¯ä»˜å¤„ç†
        return True

class WechatPayProcessor(PaymentProcessor):
    """å¾®ä¿¡æ”¯ä»˜å¤„ç†å™¨"""
    
    def process_payment(self, amount: float) -> bool:
        print(f"   å¤„ç†å¾®ä¿¡æ”¯ä»˜: Â¥{amount}")
        # æ¨¡æ‹Ÿæ”¯ä»˜å¤„ç†
        return True

# 4. é€šçŸ¥æœåŠ¡å®ç° (ISP, SRP)
class EmailNotificationService(NotificationService):
    """é‚®ä»¶é€šçŸ¥æœåŠ¡"""
    
    def send_notification(self, message: str, recipient: str) -> bool:
        print(f"   å‘é€é‚®ä»¶é€šçŸ¥åˆ° {recipient}: {message}")
        return True

class SMSNotificationService(NotificationService):
    """çŸ­ä¿¡é€šçŸ¥æœåŠ¡"""
    
    def send_notification(self, message: str, recipient: str) -> bool:
        print(f"   å‘é€çŸ­ä¿¡é€šçŸ¥åˆ° {recipient}: {message}")
        return True

# 5. è®¢å•ä»“åº“å®ç° (SRP)
class InMemoryOrderRepository(OrderRepository):
    """å†…å­˜è®¢å•ä»“åº“"""
    
    def __init__(self):
        self.orders: Dict[str, 'Order'] = {}
    
    def save_order(self, order: 'Order') -> bool:
        self.orders[order.order_id] = order
        print(f"   è®¢å• {order.order_id} å·²ä¿å­˜åˆ°å†…å­˜ä»“åº“")
        return True
    
    def get_order(self, order_id: str) -> 'Order':
        return self.orders.get(order_id)

class DatabaseOrderRepository(OrderRepository):
    """æ•°æ®åº“è®¢å•ä»“åº“"""
    
    def save_order(self, order: 'Order') -> bool:
        print(f"   è®¢å• {order.order_id} å·²ä¿å­˜åˆ°æ•°æ®åº“")
        # æ¨¡æ‹Ÿæ•°æ®åº“æ“ä½œ
        return True
    
    def get_order(self, order_id: str) -> 'Order':
        print(f"   ä»æ•°æ®åº“è·å–è®¢å• {order_id}")
        # æ¨¡æ‹Ÿæ•°æ®åº“æŸ¥è¯¢
        return None

# 6. è®¢å•ç±» (SRP, DIP)
class Order:
    """è®¢å•ç±» - å•ä¸€èŒè´£ï¼Œä¾èµ–æŠ½è±¡"""
    
    def __init__(self, order_id: str, customer_id: str):
        self.order_id = order_id
        self.customer_id = customer_id
        self.items: List[OrderItem] = []
        self.status = "pending"
        self.created_at = None
        import datetime
        self.created_at = datetime.datetime.now()
    
    def add_item(self, item: OrderItem):
        """æ·»åŠ è®¢å•é¡¹"""
        self.items.append(item)
    
    def calculate_total(self) -> float:
        """è®¡ç®—è®¢å•æ€»é¢"""
        return sum(item.calculate_price() for item in self.items)
    
    def get_description(self) -> str:
        """è·å–è®¢å•æè¿°"""
        if not self.items:
            return "ç©ºè®¢å•"
        
        items_desc = "\n".join(f"  - {item.get_description()}" for item in self.items)
        total = self.calculate_total()
        return f"è®¢å• {self.order_id}:\n{items_desc}\n  æ€»è®¡: Â¥{total}"

# 7. è®¢å•å¤„ç†æœåŠ¡ (SRP, DIP)
class OrderProcessingService:
    """è®¢å•å¤„ç†æœåŠ¡ - ä¾èµ–æ³¨å…¥"""
    
    def __init__(self, 
                 payment_processor: PaymentProcessor,
                 notification_service: NotificationService,
                 order_repository: OrderRepository):
        self.payment_processor = payment_processor
        self.notification_service = notification_service
        self.order_repository = order_repository
    
    def process_order(self, order: Order) -> bool:
        """å¤„ç†è®¢å•"""
        print(f"\n   å¼€å§‹å¤„ç†è®¢å•: {order.order_id}")
        
        # 1. è®¡ç®—æ€»é‡‘é¢
        total_amount = order.calculate_total()
        print(f"   è®¢å•æ€»é‡‘é¢: Â¥{total_amount}")
        
        # 2. å¤„ç†æ”¯ä»˜
        payment_success = self.payment_processor.process_payment(total_amount)
        
        if not payment_success:
            order.status = "payment_failed"
            print(f"   æ”¯ä»˜å¤±è´¥ï¼Œè®¢å•çŠ¶æ€: {order.status}")
            return False
        
        # 3. æ›´æ–°è®¢å•çŠ¶æ€
        order.status = "paid"
        print(f"   æ”¯ä»˜æˆåŠŸï¼Œè®¢å•çŠ¶æ€: {order.status}")
        
        # 4. ä¿å­˜è®¢å•
        save_success = self.order_repository.save_order(order)
        
        if not save_success:
            print(f"   è®¢å•ä¿å­˜å¤±è´¥")
            return False
        
        # 5. å‘é€é€šçŸ¥
        message = f"æ‚¨çš„è®¢å• {order.order_id} å·²æ”¯ä»˜æˆåŠŸï¼Œæ€»é‡‘é¢: Â¥{total_amount}"
        notification_success = self.notification_service.send_notification(
            message, f"customer_{order.customer_id}"
        )
        
        if notification_success:
            order.status = "completed"
            print(f"   é€šçŸ¥å‘é€æˆåŠŸï¼Œè®¢å•çŠ¶æ€: {order.status}")
        else:
            print(f"   é€šçŸ¥å‘é€å¤±è´¥ï¼Œä½†è®¢å•å·²å®Œæˆ")
            order.status = "completed"
        
        return True

# 8. è®¢å•å·¥å‚ (å·¥å‚æ¨¡å¼ - OCP)
class OrderFactory:
    """è®¢å•å·¥å‚ - å¯¹æ‰©å±•å¼€æ”¾"""
    
    @staticmethod
    create_product_item(product_id: str, name: str, price: float, quantity: int) -> ProductItem:
        return ProductItem(product_id, name, price, quantity)
    
    @staticmethod
    def create_service_item(service_id: str, name: str, price: float, duration: int) -> ServiceItem:
        return ServiceItem(service_id, name, price, duration)
    
    @staticmethod
    def create_order(order_id: str, customer_id: str) -> Order:
        return Order(order_id, customer_id)

# 9. ä¾èµ–æ³¨å…¥å®¹å™¨ (DIP)
class DIContainer:
    """ç®€å•çš„ä¾èµ–æ³¨å…¥å®¹å™¨"""
    
    def __init__(self):
        self._services = {}
    
    def register(self, service_type: type, implementation: type):
        self._services[service_type] = implementation
        print(f"   æ³¨å†ŒæœåŠ¡: {service_type.__name__} -> {implementation.__name__}")
    
    def get(self, service_type: type):
        if service_type in self._services:
            return self._services[service_type]()
        raise ValueError(f"æœåŠ¡ {service_type.__name__} æœªæ³¨å†Œ")
    
    def create_order_service(self, payment_type: str = "credit_card", 
                            notification_type: str = "email",
                            repository_type: str = "memory") -> OrderProcessingService:
        """åˆ›å»ºè®¢å•å¤„ç†æœåŠ¡"""
        
        # è·å–æ”¯ä»˜å¤„ç†å™¨
        if payment_type == "credit_card":
            payment_processor = CreditCardProcessor()
        elif payment_type == "alipay":
            payment_processor = AlipayProcessor()
        elif payment_type == "wechat":
            payment_processor = WechatPayProcessor()
        else:
            payment_processor = CreditCardProcessor()
        
        # è·å–é€šçŸ¥æœåŠ¡
        if notification_type == "email":
            notification_service = EmailNotificationService()
        elif notification_type == "sms":
            notification_service = SMSNotificationService()
        else:
            notification_service = EmailNotificationService()
        
        # è·å–è®¢å•ä»“åº“
        if repository_type == "memory":
            order_repository = InMemoryOrderRepository()
        elif repository_type == "database":
            order_repository = DatabaseOrderRepository()
        else:
            order_repository = InMemoryOrderRepository()
        
        return OrderProcessingService(
            payment_processor,
            notification_service,
            order_repository
        )

# 10. æµ‹è¯•ç»¼åˆåº”ç”¨
print(f"\n   æµ‹è¯•è®¾è®¡åŸåˆ™ç»¼åˆåº”ç”¨:")

# åˆ›å»ºå·¥å‚
factory = OrderFactory()

# åˆ›å»ºè®¢å•
order = factory.create_order("ORD-001", "CUST-123")

# æ·»åŠ è®¢å•é¡¹
item1 = factory.create_product_item("PROD-001", "iPhone 15", 6999.0, 1)
item2 = factory.create_product_item("PROD-002", "AirPods", 1299.0, 2)
item3 = factory.create_service_item("SERV-001", "å»¶ä¿æœåŠ¡", 299.0, 2)

order.add_item(item1)
order.add_item(item2)
order.add_item(item3)

print(f"\n     è®¢å•è¯¦æƒ…:")
print(order.get_description())

# åˆ›å»ºä¾èµ–æ³¨å…¥å®¹å™¨
container = DIContainer()

# æµ‹è¯•ä¸åŒçš„é…ç½®ç»„åˆ
configs = [
    {"payment": "credit_card", "notification": "email", "repository": "memory"},
    {"payment": "alipay", "notification": "sms", "repository": "database"},
    {"payment": "wechat", "notification": "email", "repository": "memory"}
]

for i, config in enumerate(configs, 1):
    print(f"\n     é…ç½® {i}: {config}")
    
    # åˆ›å»ºè®¢å•å¤„ç†æœåŠ¡
    order_service = container.create_order_service(
        config["payment"],
        config["notification"],
        config["repository"]
    )
    
    # å¤„ç†è®¢å•
    success = order_service.process_order(order)
    print(f"       å¤„ç†ç»“æœ: {'æˆåŠŸ' if success else 'å¤±è´¥'}")
    print(f"       æœ€ç»ˆçŠ¶æ€: {order.status}")

# 11. æ‰©å±•æ€§æµ‹è¯• (OCP)
print(f"\n     æ‰©å±•æ€§æµ‹è¯• - æ·»åŠ æ–°çš„æ”¯ä»˜æ–¹å¼:")

class BitcoinProcessor(PaymentProcessor):
    """æ¯”ç‰¹å¸æ”¯ä»˜å¤„ç†å™¨ - æ–°çš„æ”¯ä»˜æ–¹å¼"""
    
    def process_payment(self, amount: float) -> bool:
        print(f"   å¤„ç†æ¯”ç‰¹å¸æ”¯ä»˜: Â¥{amount} (çº¦{amount/400000:.8f} BTC)")
        return True

# æ‰©å±•å®¹å™¨æ”¯æŒæ¯”ç‰¹å¸æ”¯ä»˜
def create_order_service_with_bitcoin(self, notification_type="email", repository_type="memory"):
    payment_processor = BitcoinProcessor()
    
    if notification_type == "email":
        notification_service = EmailNotificationService()
    else:
        notification_service = SMSNotificationService()
    
    if repository_type == "memory":
        order_repository = InMemoryOrderRepository()
    else:
        order_repository = DatabaseOrderRepository()
    
    return OrderProcessingService(payment_processor, notification_service, order_repository)

# ç»‘å®šæ‰©å±•æ–¹æ³•
DIContainer.create_order_service_with_bitcoin = create_order_service_with_bitcoin

# æµ‹è¯•æ–°çš„æ”¯ä»˜æ–¹å¼
new_order = factory.create_order("ORD-002", "CUST-456")
new_order.add_item(factory.create_product_item("PROD-003", "MacBook Pro", 15999.0, 1))

bitcoin_service = container.create_order_service_with_bitcoin()
bitcoin_service.process_order(new_order)

print(f"\nâœ… è®¾è®¡åŸåˆ™ç»¼åˆåº”ç”¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡è®¾è®¡åŸåˆ™çš„ç»¼åˆåº”ç”¨æ–¹æ³•")
print(f"   âœ“ ç†è§£è®¾è®¡åŸåˆ™ä¹‹é—´çš„ååŒä½œç”¨")
print(f"   âœ“ èƒ½åœ¨å®é™…é¡¹ç›®ä¸­çµæ´»è¿ç”¨è®¾è®¡åŸåˆ™")
print(f"\n   ğŸ“‹ åº”ç”¨çš„è®¾è®¡åŸåˆ™æ€»ç»“:")
print(f"      â€¢ SRP: æ¯ä¸ªç±»åªè´Ÿè´£ä¸€ä¸ªèŒè´£")
print(f"      â€¢ OCP: å¯¹æ‰©å±•å¼€æ”¾ï¼Œå¯¹ä¿®æ”¹å…³é—­")
print(f"      â€¢ LSP: å­ç±»å¯ä»¥æ›¿æ¢çˆ¶ç±»")
print(f"      â€¢ ISP: æ¥å£éš”ç¦»ï¼Œå®¢æˆ·ç«¯åªä¾èµ–éœ€è¦çš„æ¥å£")
print(f"      â€¢ DIP: ä¾èµ–æŠ½è±¡è€Œä¸æ˜¯å…·ä½“å®ç°")
print(f"      â€¢ DRY: é¿å…ä»£ç é‡å¤")
print(f"      â€¢ KISS: ä¿æŒç®€å•")
print(f"      â€¢ YAGNI: åªå®ç°å½“å‰éœ€è¦çš„åŠŸèƒ½")
````
