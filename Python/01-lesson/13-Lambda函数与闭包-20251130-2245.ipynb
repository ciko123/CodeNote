### Lambdaå‡½æ•°ä¸å†…ç½®å‡½æ•°ç»“åˆ

````python
numbers = [1, 2, 3, 4]
words = ["apple", "banana", "cherry", "date", "elderberry"]

# mapå‡½æ•°
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(f"å¹³æ–¹åˆ—è¡¨: {squared_numbers}")

# filterå‡½æ•°
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"å¶æ•°åˆ—è¡¨: {even_numbers}")

# sortedå‡½æ•°çš„keyå‚æ•°
sorted_words = sorted(words, key=lambda word: len(word))
print(f"æŒ‰é•¿åº¦æ’åº: {sorted_words}")

sorted_words_reverse = sorted(words, key=lambda word: len(word), reverse=True)
print(f"æŒ‰é•¿åº¦é™åº: {sorted_words_reverse}")

# max/minå‡½æ•°çš„keyå‚æ•°
longest_word = max(words, key=lambda word: len(word))
shortest_word = min(words, key=lambda word: len(word))
print(f"æœ€é•¿å•è¯: {longest_word}")
print(f"æœ€çŸ­å•è¯: {shortest_word}")
````

### Lambdaå‡½æ•°åœ¨æ•°æ®å¤„ç†ä¸­çš„åº”ç”¨

````python
students = [
    {"name": "å¼ ä¸‰", "age": 20, "score": 85},
    {"name": "æå››", "age": 22, "score": 92},
    {"name": "ç‹äº”", "age": 19, "score": 78},
    {"name": "èµµå…­", "age": 21, "score": 95},
    {"name": "é’±ä¸ƒ", "age": 20, "score": 88}
]

# æŒ‰åˆ†æ•°æ’åº
sorted_by_score = sorted(students, key=lambda s: s["score"])
print(f"æŒ‰åˆ†æ•°æ’åº:")
for student in sorted_by_score:
    print(f"{student['name']}: {student['score']}åˆ†")

# æŒ‰å¹´é¾„æ’åº
sorted_by_age = sorted(students, key=lambda s: s["age"])
print(f"\næŒ‰å¹´é¾„æ’åº:")
for student in sorted_by_age:
    print(f"{student['name']}: {student['age']}å²")

# ç­›é€‰é«˜åˆ†å­¦ç”Ÿ
high_score_students = list(filter(lambda s: s["score"] >= 90, students))
print(f"\né«˜åˆ†å­¦ç”Ÿ(>=90åˆ†):")
for student in high_score_students:
    print(f"{student['name']}: {student['score']}åˆ†")

# æå–ç‰¹å®šå­—æ®µ
student_names = list(map(lambda s: s["name"], students))
student_ages = list(map(lambda s: s["age"], students))
print(f"\nå­¦ç”Ÿå§“å: {student_names}")
print(f"å­¦ç”Ÿå¹´é¾„: {student_ages}")
````

````python
# ä¸‰å…ƒè¿ç®—ç¬¦åœ¨lambdaä¸­
get_grade = lambda score: "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"
absolute_value = lambda x: x if x >= 0 else -x
max_of_two = lambda a, b: a if a > b else b

print(f"æˆç»©ç­‰çº§: 95->{get_grade(95)}, 85->{get_grade(85)}, 75->{get_grade(75)}, 65->{get_grade(65)}, 55->{get_grade(55)}")
print(f"ç»å¯¹å€¼: -5->{absolute_value(-5)}, 3->{absolute_value(3)}")
print(f"æœ€å¤§å€¼: max_of_two(7, 3) = {max_of_two(7, 3)}, max_of_two(2, 9) = {max_of_two(2, 9)}")

# å¤æ‚æ¡ä»¶è¡¨è¾¾å¼
categorize_number = lambda x: (
    "æ­£å¶æ•°" if x > 0 and x % 2 == 0 else
    "æ­£å¥‡æ•°" if x > 0 and x % 2 != 0 else
    "è´Ÿå¶æ•°" if x < 0 and x % 2 == 0 else
    "è´Ÿå¥‡æ•°" if x < 0 and x % 2 != 0 else
    "é›¶"
)

test_numbers = [5, -4, 0, 7, -2, 8, -3]
print(f"\næ•°å­—åˆ†ç±»:")
for num in test_numbers:
    print(f"{num}: {categorize_number(num)}")
````

### é—­åŒ…é«˜çº§åº”ç”¨ [â­â­è¿›é˜¶]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šé—­åŒ…æ˜¯å‡½æ•°å¼ç¼–ç¨‹çš„æ ¸å¿ƒæ¦‚å¿µï¼Œå…è®¸å‡½æ•°è®¿é—®å¤–éƒ¨ä½œç”¨åŸŸçš„å˜é‡ï¼Œå®ç°çŠ¶æ€å°è£…å’Œæ•°æ®éšè—ã€‚

**å­¦ä¹ è¦æ±‚**ï¼š
- æŒæ¡é—­åŒ…çš„åˆ›å»ºå’Œé«˜çº§ç”¨æ³•
- ç†è§£é—­åŒ…çš„è®¾è®¡æ¨¡å¼å’Œæœ€ä½³å®è·µ
- èƒ½åº”ç”¨é—­åŒ…è§£å†³å¤æ‚é—®é¢˜

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°å„ç§é—­åŒ…è®¾è®¡æ¨¡å¼
- è¿›è¡Œé—­åŒ…æ€§èƒ½å’Œå†…å­˜åˆ†æ
- åº”ç”¨é—­åŒ…åˆ›å»ºå®ç”¨å·¥å…·
- éªŒè¯ç‚¹ï¼šèƒ½ç‹¬ç«‹è®¾è®¡é—­åŒ…è§£å†³æ–¹æ¡ˆ

````python
print("ğŸ”— é—­åŒ…é«˜çº§åº”ç”¨:")
print("=" * 50)

# é—­åŒ…ä½œä¸ºçŠ¶æ€ç®¡ç†å™¨
print(f"ğŸ“ 1. é—­åŒ…ä½œä¸ºçŠ¶æ€ç®¡ç†å™¨:")

def create_counter(initial_value=0):
    """åˆ›å»ºè®¡æ•°å™¨é—­åŒ…"""
    count = initial_value
    
    def increment(step=1):
        nonlocal count
        count += step
        return count
    
    def decrement(step=1):
        nonlocal count
        count -= step
        return count
    
    def get_value():
        return count
    
    def reset(new_value=0):
        nonlocal count
        count = new_value
        return count
    
    def set_step(new_step):
        """è¿”å›ä¸€ä¸ªä½¿ç”¨æ–°æ­¥é•¿çš„å¢é‡å‡½æ•°"""
        return lambda: increment(new_step)
    
    return {
        'increment': increment,
        'decrement': decrement,
        'get': get_value,
        'reset': reset,
        'set_step': set_step
    }

# åˆ›å»ºè®¡æ•°å™¨å®ä¾‹
counter1 = create_counter(10)
counter2 = create_counter(0)

print(f"   è®¡æ•°å™¨1æµ‹è¯•:")
print(f"     åˆå§‹å€¼: {counter1['get']()}")
print(f"     å¢åŠ 1: {counter1['increment']()}")
print(f"     å¢åŠ 5: {counter1['increment'](5)}")
print(f"     å‡å°‘2: {counter1['decrement'](2)}")
print(f"     å½“å‰å€¼: {counter1['get']()}")

print(f"\n   è®¡æ•°å™¨2æµ‹è¯•ï¼ˆç‹¬ç«‹çŠ¶æ€ï¼‰:")
print(f"     åˆå§‹å€¼: {counter2['get']()}")
print(f"     å¢åŠ 3: {counter2['increment'](3)}")
print(f"     é‡ç½®ä¸º100: {counter2['reset'](100)}")
print(f"     å½“å‰å€¼: {counter2['get']()}")

# é—­åŒ…å®ç°ç¼“å­˜æœºåˆ¶
print(f"\nğŸ“ 2. é—­åŒ…å®ç°ç¼“å­˜æœºåˆ¶:")

def memoize_with_ttl(ttl_seconds=60):
    """åˆ›å»ºå¸¦TTLçš„è®°å¿†åŒ–è£…é¥°å™¨å·¥å‚"""
    import time
    
    cache = {}
    timestamps = {}
    
    def decorator(func):
        def memoized_func(*args):
            current_time = time.time()
            
            # æ£€æŸ¥ç¼“å­˜æ˜¯å¦å­˜åœ¨ä¸”æœªè¿‡æœŸ
            if args in cache:
                if current_time - timestamps[args] < ttl_seconds:
                    print(f"     ç¼“å­˜å‘½ä¸­: {args}")
                    return cache[args]
                else:
                    print(f"     ç¼“å­˜è¿‡æœŸ: {args}")
                    del cache[args]
                    del timestamps[args]
            
            # è®¡ç®—æ–°å€¼å¹¶ç¼“å­˜
            result = func(*args)
            cache[args] = result
            timestamps[args] = current_time
            print(f"     è®¡ç®—å¹¶ç¼“å­˜: {args} -> {result}")
            return result
        
        def clear_cache():
            cache.clear()
            timestamps.clear()
            print(f"     ç¼“å­˜å·²æ¸…ç©º")
        
        def get_cache_info():
            return {
                'size': len(cache),
                'ttl': ttl_seconds,
                'keys': list(cache.keys())
            }
        
        memoized_func.clear_cache = clear_cache
        memoized_func.cache_info = get_cache_info
        return memoized_func
    
    return decorator

# å®šä¹‰è®¡ç®—å¯†é›†çš„å‡½æ•°
@memoize_with_ttl(ttl_seconds=2)  # 2ç§’è¿‡æœŸ
def fibonacci(n):
    """è®¡ç®—æ–æ³¢é‚£å¥‘æ•°"""
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(f"   TTLç¼“å­˜æµ‹è¯•:")
print(f"     fibonacci(5): {fibonacci(5)}")
print(f"     fibonacci(5): {fibonacci(5)}")  # ä»ç¼“å­˜è·å–
print(f"     fibonacci(6): {fibonacci(6)}")
print(f"     ç¼“å­˜ä¿¡æ¯: {fibonacci.cache_info()}")

# é—­åŒ…å®ç°é…ç½®ç®¡ç†å™¨
print(f"\nğŸ“ 3. é—­åŒ…å®ç°é…ç½®ç®¡ç†å™¨:")

def create_config_manager(default_config=None):
    """åˆ›å»ºé…ç½®ç®¡ç†å™¨"""
    config = default_config.copy() if default_config else {}
    history = []  # é…ç½®å˜æ›´å†å²
    
    def get(key, default=None):
        """è·å–é…ç½®å€¼"""
        return config.get(key, default)
    
    def set(key, value):
        """è®¾ç½®é…ç½®å€¼"""
        old_value = config.get(key)
        config[key] = value
        
        # è®°å½•å˜æ›´å†å²
        history.append({
            'action': 'set',
            'key': key,
            'old_value': old_value,
            'new_value': value,
            'timestamp': len(history)
        })
        
        print(f"     é…ç½®æ›´æ–°: {key} = {value} (åŸå€¼: {old_value})")
    
    def update(**kwargs):
        """æ‰¹é‡æ›´æ–°é…ç½®"""
        for key, value in kwargs.items():
            set(key, value)
    
    def delete(key):
        """åˆ é™¤é…ç½®é¡¹"""
        if key in config:
            old_value = config.pop(key)
            history.append({
                'action': 'delete',
                'key': key,
                'old_value': old_value,
                'timestamp': len(history)
            })
            print(f"     é…ç½®åˆ é™¤: {key} (å€¼: {old_value})")
    
    def get_all():
        """è·å–æ‰€æœ‰é…ç½®"""
        return config.copy()
    
    def get_history():
        """è·å–å˜æ›´å†å²"""
        return history.copy()
    
    def reset(new_config=None):
        """é‡ç½®é…ç½®"""
        nonlocal config, history
        config = new_config.copy() if new_config else {}
        history.clear()
        print(f"     é…ç½®å·²é‡ç½®")
    
    def create_section(section_name):
        """åˆ›å»ºé…ç½®æ®µ"""
        section_config = config.get(section_name, {})
        
        def section_get(key, default=None):
            return section_config.get(key, default)
        
        def section_set(key, value):
            section_config[key] = value
            config[section_name] = section_config
            print(f"     æ®µé…ç½®æ›´æ–°: {section_name}.{key} = {value}")
        
        return {'get': section_get, 'set': section_set}
    
    return {
        'get': get,
        'set': set,
        'update': update,
        'delete': delete,
        'get_all': get_all,
        'history': get_history,
        'reset': reset,
        'create_section': create_section
    }

# åˆ›å»ºé…ç½®ç®¡ç†å™¨
config_mgr = create_config_manager({
    'debug': False,
    'version': '1.0.0',
    'timeout': 30
})

print(f"   é…ç½®ç®¡ç†å™¨æµ‹è¯•:")
print(f"     åˆå§‹é…ç½®: {config_mgr['get_all']()}")

config_mgr['set']('debug', True)
config_mgr['set']('timeout', 60)
config_mgr['update'](author='AI Assistant', license='MIT')

print(f"\n     æ›´æ–°åé…ç½®: {config_mgr['get_all']()}")
print(f"     é…ç½®å†å²: {config_mgr['history']()}")

# åˆ›å»ºé…ç½®æ®µ
database_config = config_mgr['create_section']('database')
database_config['set']('host', 'localhost')
database_config['set']('port', 5432)

print(f"\n     æ®µé…ç½®å: {config_mgr['get_all']()}")

# é—­åŒ…å®ç°äº‹ä»¶ç³»ç»Ÿ
print(f"\nğŸ“ 4. é—­åŒ…å®ç°äº‹ä»¶ç³»ç»Ÿ:")

def create_event_emitter():
    """åˆ›å»ºäº‹ä»¶å‘å°„å™¨"""
    listeners = {}
    event_log = []
    
    def on(event_name, callback):
        """æ³¨å†Œäº‹ä»¶ç›‘å¬å™¨"""
        if event_name not in listeners:
            listeners[event_name] = []
        listeners[event_name].append(callback)
        print(f"     æ³¨å†Œç›‘å¬å™¨: {event_name}")
    
    def emit(event_name, *args, **kwargs):
        """å‘å°„äº‹ä»¶"""
        event_data = {
            'name': event_name,
            'args': args,
            'kwargs': kwargs,
            'timestamp': len(event_log)
        }
        event_log.append(event_data)
        
        print(f"     å‘å°„äº‹ä»¶: {event_name}")
        
        if event_name in listeners:
            for callback in listeners[event_name]:
                try:
                    callback(*args, **kwargs)
                except Exception as e:
                    print(f"       ç›‘å¬å™¨é”™è¯¯: {e}")
    
    def off(event_name, callback=None):
        """ç§»é™¤äº‹ä»¶ç›‘å¬å™¨"""
        if event_name in listeners:
            if callback:
                if callback in listeners[event_name]:
                    listeners[event_name].remove(callback)
                    print(f"     ç§»é™¤ç‰¹å®šç›‘å¬å™¨: {event_name}")
            else:
                listeners[event_name].clear()
                print(f"     ç§»é™¤æ‰€æœ‰ç›‘å¬å™¨: {event_name}")
    
    def once(event_name, callback):
        """æ³¨å†Œä¸€æ¬¡æ€§äº‹ä»¶ç›‘å¬å™¨"""
        def once_callback(*args, **kwargs):
            callback(*args, **kwargs)
            off(event_name, once_callback)
        
        on(event_name, once_callback)
        print(f"     æ³¨å†Œä¸€æ¬¡æ€§ç›‘å¬å™¨: {event_name}")
    
    def get_event_log():
        return event_log.copy()
    
    def clear_log():
        event_log.clear()
    
    return {
        'on': on,
        'emit': emit,
        'off': off,
        'once': once,
        'log': get_event_log,
        'clear_log': clear_log
    }

# åˆ›å»ºäº‹ä»¶ç³»ç»Ÿ
emitter = create_event_emitter()

# å®šä¹‰äº‹ä»¶å¤„ç†å‡½æ•°
def on_user_login(username):
    print(f"       ç”¨æˆ·ç™»å½•: {username}")

def on_user_logout(username):
    print(f"       ç”¨æˆ·ç™»å‡º: {username}")

def on_data_change(data_type, count):
    print(f"       æ•°æ®å˜æ›´: {data_type}, æ•°é‡: {count}")

print(f"   äº‹ä»¶ç³»ç»Ÿæµ‹è¯•:")

# æ³¨å†Œç›‘å¬å™¨
emitter['on']('login', on_user_login)
emitter['on']('logout', on_user_logout)
emitter['on']('data_change', on_data_change)

# å‘å°„äº‹ä»¶
emitter['emit']('login', 'å¼ ä¸‰')
emitter['emit']('data_change', 'ç”¨æˆ·æ•°æ®', 100)
emitter['emit']('logout', 'å¼ ä¸‰')

# ä¸€æ¬¡æ€§ç›‘å¬å™¨
emitter['once']('special_event', lambda: print("       ç‰¹æ®Šäº‹ä»¶ï¼ˆä»…ä¸€æ¬¡ï¼‰"))
emitter['emit']('special_event')
emitter['emit']('special_event')  # ä¸ä¼šè§¦å‘

print(f"\n     äº‹ä»¶æ—¥å¿—: {emitter['log']()}")

print(f"\nâœ… é—­åŒ…é«˜çº§åº”ç”¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡é—­åŒ…é«˜çº§ç”¨æ³•")
print(f"   âœ“ ç†è§£é—­åŒ…è®¾è®¡æ¨¡å¼")
print(f"   âœ“ èƒ½åº”ç”¨é—­åŒ…è§£å†³å¤æ‚é—®é¢˜")
````

````python
print("ğŸ”§ å‡½æ•°å¼ç¼–ç¨‹æ¨¡å¼:")
print("=" * 50)

# é«˜é˜¶å‡½æ•°
print(f"ğŸ“ 1. é«˜é˜¶å‡½æ•°:")

def apply_twice(func, x):
    """åº”ç”¨å‡½æ•°ä¸¤æ¬¡"""
    return func(func(x))

def compose(f, g):
    """å‡½æ•°ç»„åˆ"""
    return lambda x: f(g(x))

def pipe(*functions):
    """å‡½æ•°ç®¡é“"""
    return lambda x: reduce(lambda acc, f: f(acc), functions, x)

from functools import reduce

# æµ‹è¯•é«˜é˜¶å‡½æ•°
square = lambda x: x ** 2
increment = lambda x: x + 1
halve = lambda x: x / 2

print(f"   apply_twice(square, 3) = {apply_twice(square, 3)}")
print(f"   apply_twice(increment, 5) = {apply_twice(increment, 5)}")

# å‡½æ•°ç»„åˆ
square_then_increment = compose(increment, square)
increment_then_square = compose(square, increment)

print(f"\n   square_then_increment(3) = {square_then_increment(3)}")
print(f"   increment_then_square(3) = {increment_then_square(3)}")

# å‡½æ•°ç®¡é“
pipeline = pipe(increment, square, halve)
print(f"   pipeline(3) = {pipeline(3)}")  # ((3 + 1) ** 2) / 2 = 8

# æŸ¯é‡ŒåŒ–ï¼ˆCurryingï¼‰
print(f"\nğŸ“ 2. æŸ¯é‡ŒåŒ–ï¼ˆCurryingï¼‰:")

def curry(func):
    """æŸ¯é‡ŒåŒ–è£…é¥°å™¨"""
    def curried(*args, **kwargs):
        if len(args) + len(kwargs) >= func.__code__.co_argcount:
            return func(*args, **kwargs)
        else:
            return lambda *more_args, **more_kwargs: curried(*(args + more_args), **{**kwargs, **more_kwargs})
    return curried

# æ™®é€šå‡½æ•°
def add_three(a, b, c):
    return a + b + c

# æŸ¯é‡ŒåŒ–ç‰ˆæœ¬
@curry
def curried_add_three(a, b, c):
    return a + b + c

print(f"   æ™®é€šå‡½æ•°: add_three(1, 2, 3) = {add_three(1, 2, 3)}")
print(f"   æŸ¯é‡ŒåŒ–é€æ­¥è°ƒç”¨:")

step1 = curried_add_three(1)
step2 = step1(2)
result = step2(3)
print(f"     step1 = curried_add_three(1): {step1}")
print(f"     step2 = step1(2): {step2}")
print(f"     result = step2(3): {result}")

# ä¹Ÿå¯ä»¥ä¸€æ¬¡æ€§è°ƒç”¨
print(f"     ä¸€æ¬¡æ€§è°ƒç”¨: curried_add_three(1, 2, 3) = {curried_add_three(1, 2, 3)}")

# åå‡½æ•°åº”ç”¨
print(f"\nğŸ“ 3. åå‡½æ•°åº”ç”¨:")
from functools import partial

def power(base, exponent):
    """è®¡ç®—å¹‚"""
    return base ** exponent

# åˆ›å»ºåå‡½æ•°
square = partial(power, exponent=2)
cube = partial(power, exponent=3)
power_of_two = partial(power, base=2)

print(f"   square(5) = {square(5)}")
print(f"   cube(3) = {cube(3)}")
print(f"   power_of_two(5) = {power_of_two(5)}")

# å®é™…åº”ç”¨ï¼šå­—ç¬¦ä¸²å¤„ç†
def format_string(template, *args, **kwargs):
    """æ ¼å¼åŒ–å­—ç¬¦ä¸²"""
    return template.format(*args, **kwargs)

# åˆ›å»ºä¸“é—¨çš„æ ¼å¼åŒ–å‡½æ•°
greet = partial(format_string, "Hello, {}!")
log_info = partial(format_string, "[INFO] {} - {}")
log_error = partial(format_string, "[ERROR] {} - {}")

print(f"\n   greet('å¼ ä¸‰') = {greet('å¼ ä¸‰')}")
print(f"   log_info('App', 'Started') = {log_info('App', 'Started')}")
print(f"   log_error('DB', 'Connection failed') = {log_error('DB', 'Connection failed')}")

# çº¯å‡½æ•°å’Œä¸å¯å˜æ€§
print(f"\nğŸ“ 4. çº¯å‡½æ•°å’Œä¸å¯å˜æ€§:")

# éçº¯å‡½æ•°ï¼ˆæœ‰å‰¯ä½œç”¨ï¼‰
def impure_add_to_list(lst, item):
    lst.append(item)  # ä¿®æ”¹åŸå§‹åˆ—è¡¨
    return lst

# çº¯å‡½æ•°ï¼ˆæ— å‰¯ä½œç”¨ï¼‰
def pure_add_to_list(lst, item):
    return lst + [item]  # è¿”å›æ–°åˆ—è¡¨

# ä¸å¯å˜æ•°æ®ç»“æ„æ“ä½œ
def immutable_update(obj, **updates):
    """ä¸å¯å˜æ›´æ–°å¯¹è±¡"""
    if isinstance(obj, dict):
        new_obj = obj.copy()
        new_obj.update(updates)
        return new_obj
    elif hasattr(obj, '_replace'):  # namedtupleç­‰
        return obj._replace(**updates)
    else:
        raise ValueError("Unsupported type for immutable update")

# æµ‹è¯•çº¯å‡½æ•°
original_list = [1, 2, 3]
result1 = impure_add_to_list(original_list, 4)
print(f"   éçº¯å‡½æ•°: åŸå§‹={original_list}, ç»“æœ={result1}")

original_list = [1, 2, 3]
result2 = pure_add_to_list(original_list, 4)
print(f"   çº¯å‡½æ•°: åŸå§‹={original_list}, ç»“æœ={result2}")

# ä¸å¯å˜æ›´æ–°
original_dict = {'name': 'å¼ ä¸‰', 'age': 25}
updated_dict = immutable_update(original_dict, age=26, city='åŒ—äº¬')
print(f"\n   ä¸å¯å˜æ›´æ–°: åŸå§‹={original_dict}, æ›´æ–°å={updated_dict}")

# å‡½æ•°å¼æ•°æ®å¤„ç†
print(f"\nğŸ“ 5. å‡½æ•°å¼æ•°æ®å¤„ç†:")

def functional_filter(predicate, data):
    """å‡½æ•°å¼è¿‡æ»¤"""
    return [item for item in data if predicate(item)]

def functional_map(transform, data):
    """å‡½æ•°å¼æ˜ å°„"""
    return [transform(item) for item in data]

def functional_reduce(combiner, data, initial=None):
    """å‡½æ•°å¼å½’çº¦"""
    if initial is not None:
        return reduce(combiner, data, initial)
    else:
        return reduce(combiner, data)

# æ•°æ®å¤„ç†ç®¡é“
def create_processing_pipeline(*operations):
    """åˆ›å»ºæ•°æ®å¤„ç†ç®¡é“"""
    def pipeline(data):
        result = data
        for operation in operations:
            result = operation(result)
        return result
    return pipeline

# æµ‹è¯•æ•°æ®å¤„ç†
numbers = list(range(1, 11))
print(f"   åŸå§‹æ•°æ®: {numbers}")

# å‡½æ•°å¼æ“ä½œé“¾
evens = functional_filter(lambda x: x % 2 == 0, numbers)
squared = functional_map(lambda x: x ** 2, evens)
sum_of_squares = functional_reduce(lambda a, b: a + b, squared)

print(f"   å¶æ•°: {evens}")
print(f"   å¹³æ–¹: {squared}")
print(f"   å¹³æ–¹å’Œ: {sum_of_squares}")

# ä½¿ç”¨ç®¡é“
pipeline = create_processing_pipeline(
    lambda data: functional_filter(lambda x: x % 2 == 0, data),
    lambda data: functional_map(lambda x: x ** 2, data),
    lambda data: functional_reduce(lambda a, b: a + b, data)
)

pipeline_result = pipeline(numbers)
print(f"\n   ç®¡é“å¤„ç†ç»“æœ: {pipeline_result}")

# å‡½æ•°å¼å·¥å…·å‡½æ•°
print(f"\nğŸ“ 6. å‡½æ•°å¼å·¥å…·å‡½æ•°:")

def compose_multiple(*functions):
    """ç»„åˆå¤šä¸ªå‡½æ•°"""
    return reduce(lambda f, g: lambda x: f(g(x)), functions, lambda x: x)

def memoize(func):
    """è®°å¿†åŒ–è£…é¥°å™¨"""
    cache = {}
    def memoized(*args):
        if args not in cache:
            cache[args] = func(*args)
        return cache[args]
    return memoized

def lazy_map(func, data):
    """æƒ°æ€§æ˜ å°„ï¼ˆç”Ÿæˆå™¨ç‰ˆæœ¬ï¼‰"""
    return (func(item) for item in data)

def lazy_filter(predicate, data):
    """æƒ°æ€§è¿‡æ»¤ï¼ˆç”Ÿæˆå™¨ç‰ˆæœ¬ï¼‰"""
    return (item for item in data if predicate(item))

# æµ‹è¯•å·¥å…·å‡½æ•°
add_1 = lambda x: x + 1
multiply_2 = lambda x: x * 2
subtract_3 = lambda x: x - 3

composed = compose_multiple(subtract_3, multiply_2, add_1)
print(f"   ç»„åˆå‡½æ•°: ((5 + 1) * 2) - 3 = {composed(5)}")

# è®°å¿†åŒ–æ–æ³¢é‚£å¥‘
@memoize
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(f"\n   è®°å¿†åŒ–æ–æ³¢é‚£å¥‘: fib(10) = {fib(10)}")

# æƒ°æ€§å¤„ç†
lazy_result = lazy_map(lambda x: x ** 2, lazy_filter(lambda x: x % 2 == 0, range(1, 11)))
print(f"   æƒ°æ€§å¤„ç†ç»“æœ: {list(lazy_result)}")

print(f"\nâœ… å‡½æ•°å¼ç¼–ç¨‹æ¨¡å¼å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡å‡½æ•°å¼ç¼–ç¨‹æ ¸å¿ƒæ¦‚å¿µ")
print(f"   âœ“ ç†è§£é«˜é˜¶å‡½æ•°è®¾è®¡åº”ç”¨")
print(f"   âœ“ èƒ½åº”ç”¨å‡½æ•°å¼ç¼–ç¨‹èŒƒå¼")
````

````python
print("ğŸ’¼ å®é™…åº”ç”¨æ¡ˆä¾‹:")
print("=" * 50)

# æ¡ˆä¾‹1ï¼šå‡½æ•°å¼æ•°æ®æŸ¥è¯¢æ„å»ºå™¨
print(f"ğŸ“ 1. å‡½æ•°å¼æ•°æ®æŸ¥è¯¢æ„å»ºå™¨:")

def create_query_builder():
    """åˆ›å»ºæŸ¥è¯¢æ„å»ºå™¨"""
    filters = []
    sorters = []
    limit_count = None
    
    def where(predicate):
        """æ·»åŠ è¿‡æ»¤æ¡ä»¶"""
        filters.append(predicate)
        return QueryBuilder(filters, sorters, limit_count)
    
    def order_by(key_func, reverse=False):
        """æ·»åŠ æ’åºæ¡ä»¶"""
        sorters.append((key_func, reverse))
        return QueryBuilder(filters, sorters, limit_count)
    
    def limit(count):
        """è®¾ç½®é™åˆ¶æ•°é‡"""
        nonlocal limit_count
        limit_count = count
        return QueryBuilder(filters, sorters, limit_count)
    
    def execute(data):
        """æ‰§è¡ŒæŸ¥è¯¢"""
        result = data.copy()
        
        # åº”ç”¨è¿‡æ»¤å™¨
        for filter_func in filters:
            result = [item for item in result if filter_func(item)]
        
        # åº”ç”¨æ’åºå™¨
        for key_func, reverse in sorters:
            result = sorted(result, key=key_func, reverse=reverse)
        
        # åº”ç”¨é™åˆ¶
        if limit_count:
            result = result[:limit_count]
        
        return result
    
    class QueryBuilder:
        def __init__(self, filters, sorters, limit_count):
            self.filters = filters
            self.sorters = sorters
            self.limit_count = limit_count
        
        def where(self, predicate):
            new_filters = self.filters + [predicate]
            return QueryBuilder(new_filters, self.sorters, self.limit_count)
        
        def order_by(self, key_func, reverse=False):
            new_sorters = self.sorters + [(key_func, reverse)]
            return QueryBuilder(self.filters, new_sorters, self.limit_count)
        
        def limit(self, count):
            return QueryBuilder(self.filters, self.sorters, count)
        
        def execute(self, data):
            result = data.copy()
            
            for filter_func in self.filters:
                result = [item for item in result if filter_func(item)]
            
            for key_func, reverse in self.sorters:
                result = sorted(result, key=key_func, reverse=reverse)
            
            if self.limit_count:
                result = result[:self.limit_count]
            
            return result
    
    return QueryBuilder([], [], None)

# æµ‹è¯•æŸ¥è¯¢æ„å»ºå™¨
users = [
    {"name": "å¼ ä¸‰", "age": 25, "score": 85, "city": "åŒ—äº¬"},
    {"name": "æå››", "age": 30, "score": 92, "city": "ä¸Šæµ·"},
    {"name": "ç‹äº”", "age": 22, "score": 78, "city": "å¹¿å·"},
    {"name": "èµµå…­", "age": 28, "score": 95, "city": "æ·±åœ³"},
    {"name": "é’±ä¸ƒ", "age": 35, "score": 88, "city": "åŒ—äº¬"},
    {"name": "å­™å…«", "age": 27, "score": 82, "city": "ä¸Šæµ·"}
]

query = create_query_builder()

print(f"   æŸ¥è¯¢æ„å»ºå™¨æµ‹è¯•:")

# å¤æ‚æŸ¥è¯¢ï¼šå¹´é¾„å¤§äº25ï¼Œåˆ†æ•°å¤§äº80ï¼ŒæŒ‰åˆ†æ•°é™åºï¼Œå–å‰3å
result1 = query.where(lambda u: u["age"] > 25)\
                  .where(lambda u: u["score"] > 80)\
                  .order_by(lambda u: u["score"], reverse=True)\
                  .limit(3)\
                  .execute(users)

print(f"     å¤æ‚æŸ¥è¯¢ç»“æœ:")
for user in result1:
    print(f"       {user['name']}: {user['age']}å², {user['score']}åˆ†")

# ç®€å•æŸ¥è¯¢ï¼šæ‰€æœ‰åŒ—äº¬ç”¨æˆ·ï¼ŒæŒ‰å¹´é¾„æ’åº
result2 = query.where(lambda u: u["city"] == "åŒ—äº¬")\
                  .order_by(lambda u: u["age"])\
                  .execute(users)

print(f"\n     åŒ—äº¬ç”¨æˆ·:")
for user in result2:
    print(f"       {user['name']}: {user['age']}å²")

# æ¡ˆä¾‹2ï¼šå‡½æ•°å¼çŠ¶æ€ç®¡ç†å™¨
print(f"\nğŸ“ 2. å‡½æ•°å¼çŠ¶æ€ç®¡ç†å™¨:")

def create_reducer(initial_state, actions):
    """åˆ›å»ºReduxé£æ ¼çš„reducer"""
    state = initial_state
    
    def dispatch(action_type, payload=None):
        nonlocal state
        
        if action_type in actions:
            old_state = state
            state = actions[action_type](state, payload)
            print(f"     Action: {action_type}, Payload: {payload}")
            print(f"       State: {old_state} -> {state}")
            return state
        else:
            print(f"     Unknown action: {action_type}")
            return state
    
    def get_state():
        return state
    
    def subscribe(listener):
        """è®¢é˜…çŠ¶æ€å˜åŒ–"""
        def unsubscribe():
            pass  # ç®€åŒ–å®ç°
        return unsubscribe
    
    return {
        'dispatch': dispatch,
        'get_state': get_state,
        'subscribe': subscribe
    }

# å®šä¹‰è®¡æ•°å™¨çš„actions
counter_actions = {
    'INCREMENT': lambda state, payload: state + (payload or 1),
    'DECREMENT': lambda state, payload: state - (payload or 1),
    'RESET': lambda state, payload: 0,
    'SET': lambda state, payload: payload if payload is not None else state
}

# åˆ›å»ºè®¡æ•°å™¨store
counter_store = create_reducer(0, counter_actions)

print(f"   å‡½æ•°å¼çŠ¶æ€ç®¡ç†æµ‹è¯•:")
print(f"     åˆå§‹çŠ¶æ€: {counter_store['get_state']()}")

counter_store['dispatch']('INCREMENT')
counter_store['dispatch']('INCREMENT', 5)
counter_store['dispatch']('DECREMENT', 2)
counter_store['dispatch']('RESET')
counter_store['dispatch']('SET', 100)

# å®šä¹‰è´­ç‰©è½¦çš„actions
cart_actions = {
    'ADD_ITEM': lambda state, payload: {
        **state,
        'items': state['items'] + [payload],
        'total': state['total'] + payload['price']
    },
    'REMOVE_ITEM': lambda state, payload: {
        **state,
        'items': [item for item in state['items'] if item['id'] != payload],
        'total': sum(item['price'] for item in state['items'] if item['id'] != payload)
    },
    'CLEAR_CART': lambda state, payload: {
        **state,
        'items': [],
        'total': 0
    }
}

# åˆ›å»ºè´­ç‰©è½¦store
initial_cart_state = {'items': [], 'total': 0}
cart_store = create_reducer(initial_cart_state, cart_actions)

print(f"\n   è´­ç‰©è½¦çŠ¶æ€ç®¡ç†:")
print(f"     åˆå§‹çŠ¶æ€: {cart_store['get_state']()}")

cart_store['dispatch']('ADD_ITEM', {'id': 1, 'name': 'è‹¹æœ', 'price': 5})
cart_store['dispatch']('ADD_ITEM', {'id': 2, 'name': 'é¦™è•‰', 'price': 3})
cart_store['dispatch']('ADD_ITEM', {'id': 3, 'name': 'æ©™å­', 'price': 4})
cart_store['dispatch']('REMOVE_ITEM', 2)
cart_store['dispatch']('CLEAR_CART')

# æ¡ˆä¾‹3ï¼šå‡½æ•°å¼ä¸­é—´ä»¶ç³»ç»Ÿ
print(f"\nğŸ“ 3. å‡½æ•°å¼ä¸­é—´ä»¶ç³»ç»Ÿ:")

def create_middleware_system():
    """åˆ›å»ºä¸­é—´ä»¶ç³»ç»Ÿ"""
    middlewares = []
    
    def use(middleware):
        """æ·»åŠ ä¸­é—´ä»¶"""
        middlewares.append(middleware)
        print(f"     æ·»åŠ ä¸­é—´ä»¶: {middleware.__name__}")
    
    def create_handler(final_handler):
        """åˆ›å»ºå¤„ç†å‡½æ•°"""
        def handler(request):
            # æ„å»ºä¸­é—´ä»¶é“¾
            chain = final_handler
            
            for middleware in reversed(middlewares):
                def create_next(current_chain, current_middleware):
                    def next_handler(req):
                        return current_middleware(req, current_chain)
                    return next_handler
                chain = create_next(chain, middleware)
            
            return chain(request)
        
        return handler
    
    return {
        'use': use,
        'create_handler': create_handler
    }

# å®šä¹‰ä¸­é—´ä»¶
def logging_middleware(request, next_handler):
    """æ—¥å¿—ä¸­é—´ä»¶"""
    print(f"       [æ—¥å¿—] è¯·æ±‚å¼€å§‹: {request}")
    response = next_handler(request)
    print(f"       [æ—¥å¿—] è¯·æ±‚ç»“æŸ: {response}")
    return response

def auth_middleware(request, next_handler):
    """è®¤è¯ä¸­é—´ä»¶"""
    if request.get('authenticated', False):
        print(f"       [è®¤è¯] ç”¨æˆ·å·²è®¤è¯")
        return next_handler(request)
    else:
        print(f"       [è®¤è¯] ç”¨æˆ·æœªè®¤è¯")
        return {'error': 'Unauthorized', 'status': 401}

def validation_middleware(request, next_handler):
    """éªŒè¯ä¸­é—´ä»¶"""
    if 'data' in request and isinstance(request['data'], dict):
        print(f"       [éªŒè¯] æ•°æ®æ ¼å¼æ­£ç¡®")
        return next_handler(request)
    else:
        print(f"       [éªŒè¯] æ•°æ®æ ¼å¼é”™è¯¯")
        return {'error': 'Invalid data', 'status': 400}

# åˆ›å»ºä¸­é—´ä»¶ç³»ç»Ÿ
middleware_system = create_middleware_system()

print(f"   ä¸­é—´ä»¶ç³»ç»Ÿæµ‹è¯•:")

# æ·»åŠ ä¸­é—´ä»¶
middleware_system['use'](logging_middleware)
middleware_system['use'](auth_middleware)
middleware_system['use'](validation_middleware)

# å®šä¹‰æœ€ç»ˆå¤„ç†å™¨
def final_handler(request):
    """æœ€ç»ˆä¸šåŠ¡å¤„ç†å™¨"""
    print(f"       [ä¸šåŠ¡] å¤„ç†ä¸šåŠ¡é€»è¾‘")
    return {'success': True, 'data': request['data'], 'status': 200}

# åˆ›å»ºå¤„ç†å‡½æ•°
handler = middleware_system['create_handler'](final_handler)

# æµ‹è¯•è¯·æ±‚
print(f"\n     æµ‹è¯•è®¤è¯è¯·æ±‚:")
auth_request = {
    'authenticated': True,
    'data': {'name': 'å¼ ä¸‰', 'age': 25}
}
response1 = handler(auth_request)
print(f"       å“åº”: {response1}")

print(f"\n     æµ‹è¯•æœªè®¤è¯è¯·æ±‚:")
unauth_request = {
    'authenticated': False,
    'data': {'name': 'æå››', 'age': 30}
}
response2 = handler(unauth_request)
print(f"       å“åº”: {response2}")

print(f"\n     æµ‹è¯•æ— æ•ˆæ•°æ®è¯·æ±‚:")
invalid_request = {
    'authenticated': True,
    'data': 'invalid data'
}
response3 = handler(invalid_request)
print(f"       å“åº”: {response3}")

# æ¡ˆä¾‹4ï¼šå‡½æ•°å¼å·¥å…·åº“
print(f"\nğŸ“ 4. å‡½æ•°å¼å·¥å…·åº“:")

class FunctionalUtils:
    @staticmethod
    def compose(*functions):
        """å‡½æ•°ç»„åˆ"""
        return reduce(lambda f, g: lambda x: f(g(x)), functions, lambda x: x)
    
    @staticmethod
    def pipe(*functions):
        """å‡½æ•°ç®¡é“"""
        return reduce(lambda f, g: lambda x: g(f(x)), functions, lambda x: x)
    
    @staticmethod
    def memoize(func):
        """è®°å¿†åŒ–"""
        cache = {}
        def memoized(*args):
            if args not in cache:
                cache[args] = func(*args)
            return cache[args]
        return memoized
    
    @staticmethod
    def curry(func):
        """æŸ¯é‡ŒåŒ–"""
        def curried(*args, **kwargs):
            if len(args) + len(kwargs) >= func.__code__.co_argcount:
                return func(*args, **kwargs)
            else:
                return lambda *more_args, **more_kwargs: curried(*(args + more_args), **{**kwargs, **more_kwargs})
        return curried
    
    @staticmethod
    def partial(func, *args, **kwargs):
        """åå‡½æ•°"""
        def partial_func(*more_args, **more_kwargs):
            return func(*(args + more_args), **{**kwargs, **more_kwargs})
        return partial_func
    
    @staticmethod
    def lazy_map(func, iterable):
        """æƒ°æ€§æ˜ å°„"""
        return (func(item) for item in iterable)
    
    @staticmethod
    def lazy_filter(predicate, iterable):
        """æƒ°æ€§è¿‡æ»¤"""
        return (item for item in iterable if predicate(item))
    
    @staticmethod
    def group_by(key_func, data):
        """åˆ†ç»„"""
        groups = {}
        for item in data:
            key = key_func(item)
            if key not in groups:
                groups[key] = []
            groups[key].append(item)
        return groups

# æµ‹è¯•å‡½æ•°å¼å·¥å…·åº“
print(f"   å‡½æ•°å¼å·¥å…·åº“æµ‹è¯•:")

# ç»„åˆå‡½æ•°
add_and_square = FunctionalUtils.compose(lambda x: x ** 2, lambda x: x + 1)
print(f"     ç»„åˆå‡½æ•°: add_and_square(4) = {add_and_square(4)}")

# ç®¡é“å‡½æ•°
pipeline = FunctionalUtils.pipe(
    lambda x: x + 1,
    lambda x: x * 2,
    lambda x: x - 3
)
print(f"     ç®¡é“å‡½æ•°: pipeline(5) = {pipeline(5)}")

# è®°å¿†åŒ–
@FunctionalUtils.memoize
def expensive_calculation(n):
    print(f"       æ‰§è¡Œå¤æ‚è®¡ç®—: {n}")
    return sum(range(n + 1))

print(f"\n     è®°å¿†åŒ–æµ‹è¯•:")
print(f"       ç¬¬ä¸€æ¬¡: {expensive_calculation(100)}")
print(f"       ç¬¬äºŒæ¬¡: {expensive_calculation(100)}")

# åˆ†ç»„
people = [
    {'name': 'å¼ ä¸‰', 'age': 25, 'city': 'åŒ—äº¬'},
    {'name': 'æå››', 'age': 30, 'city': 'ä¸Šæµ·'},
    {'name': 'ç‹äº”', 'age': 25, 'city': 'åŒ—äº¬'},
    {'name': 'èµµå…­', 'age': 30, 'city': 'å¹¿å·'}
]

grouped_by_age = FunctionalUtils.group_by(lambda p: p['age'], people)
grouped_by_city = FunctionalUtils.group_by(lambda p: p['city'], people)

print(f"\n     æŒ‰å¹´é¾„åˆ†ç»„:")
for age, group in grouped_by_age.items():
    names = [p['name'] for p in group]
    print(f"       {age}å²: {names}")

print(f"\n     æŒ‰åŸå¸‚åˆ†ç»„:")
for city, group in grouped_by_city.items():
    names = [p['name'] for p in group]
    print(f"       {city}: {names}")

print(f"\nâœ… å®é™…åº”ç”¨æ¡ˆä¾‹å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡Lambdaå’Œé—­åŒ…å®é™…åº”ç”¨")
print(f"   âœ“ ç†è§£å‡½æ•°å¼ç¼–ç¨‹è®¾è®¡æ¨¡å¼")
print(f"   âœ“ èƒ½è®¾è®¡ä¼˜é›…å‡½æ•°å¼è§£å†³æ–¹æ¡ˆ")
````
