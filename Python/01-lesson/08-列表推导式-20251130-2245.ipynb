### åµŒå¥—å’Œé«˜çº§æ¨å¯¼å¼ [â­â­è¿›é˜¶]
**çŸ¥è¯†ç‚¹è¯´æ˜**ï¼šåµŒå¥—æ¨å¯¼å¼å’Œé«˜çº§æŠ€å·§å¯ä»¥å¤„ç†æ›´å¤æ‚çš„æ•°æ®ç»“æ„å’Œç®—æ³•é—®é¢˜ã€‚

- æŒæ¡åµŒå¥—åˆ—è¡¨æ¨å¯¼å¼

- èƒ½è¿›è¡Œå¤æ‚æ•°æ®ç»“æ„å¤„ç†

**æ¡ˆä¾‹è¦æ±‚**ï¼š
- å®ç°åµŒå¥—æ¨å¯¼å¼
- è¿›è¡Œé«˜çº§æ¨å¯¼å¼åº”ç”¨
- åº”ç”¨æ¨å¯¼å¼è§£å†³å¤æ‚é—®é¢˜
- éªŒè¯ç‚¹ï¼šèƒ½æ­£ç¡®ä½¿ç”¨åµŒå¥—å’Œé«˜çº§æ¨å¯¼å¼

````python
print("ğŸ¯ æ¨å¯¼å¼å®é™…åº”ç”¨:")
print("=" * 50)

# æ¡ˆä¾‹1ï¼šæ•°æ®åˆ†æç®¡é“
print(f"ğŸ“ 1. æ•°æ®åˆ†æç®¡é“:")

def data_analysis_pipeline(raw_data):
    """æ•°æ®åˆ†æç®¡é“"""
    print(f"   åŸå§‹æ•°æ®: {raw_data}")
    
    # æ­¥éª¤1: æ•°æ®æ¸…æ´—
    cleaned = [x for x in raw_data if isinstance(x, (int, float)) and not str(x).lower() == 'nan']
    print(f"   æ¸…æ´—å: {cleaned}")
    
    # æ­¥éª¤2: å¼‚å¸¸å€¼æ£€æµ‹å’Œç§»é™¤
    if cleaned:
        mean_val = sum(cleaned) / len(cleaned)
        std_val = (sum((x - mean_val) ** 2 for x in cleaned) / len(cleaned)) ** 0.5
        
        # ç§»é™¤è¶…å‡º3å€æ ‡å‡†å·®çš„å€¼
        filtered = [x for x in cleaned if abs(x - mean_val) <= 3 * std_val]
        print(f"   å¼‚å¸¸å€¼è¿‡æ»¤å: {filtered}")
    else:
        filtered = cleaned
    
    # æ­¥éª¤3: æ•°æ®è½¬æ¢
    normalized = [(x - min(filtered)) / (max(filtered) - min(filtered)) if filtered else 0 
                  for x in filtered]
    print(f"   æ ‡å‡†åŒ–å: {[f'{x:.3f}' for x in normalized]}")
    
    # æ­¥éª¤4: æ•°æ®åˆ†ç±»
    categories = [
        "é«˜" if x > 0.7 else "ä¸­" if x > 0.3 else "ä½"
        for x in normalized
    ]
    print(f"   åˆ†ç±»ç»“æœ: {categories}")
    
    # æ­¥éª¤5: ç»Ÿè®¡åˆ†æ
    if filtered:
        stats = {
            'count': len(filtered),
            'mean': sum(filtered) / len(filtered),
            'median': sorted(filtered)[len(filtered) // 2],
            'min': min(filtered),
            'max': max(filtered)
        }
        print(f"\n   ç»Ÿè®¡ä¿¡æ¯:")
        for key, value in stats.items():
            print(f"     {key}: {value:.3f}")
    
    return {
        'original': raw_data,
        'cleaned': cleaned,
        'filtered': filtered,
        'normalized': normalized,
        'categories': categories
    }

# æµ‹è¯•æ•°æ®åˆ†æ
test_data = [10, 25, 30, 45, 60, 75, 90, 100, 15, 35, 55, 80, 200, 5, 95]
analysis_result = data_analysis_pipeline(test_data)

# æ¡ˆä¾‹2ï¼šæ–‡æœ¬å¤„ç†å’Œåˆ†æ
print(f"\nğŸ“ 2. æ–‡æœ¬å¤„ç†å’Œåˆ†æ:")

def text_analysis(text):
    """æ–‡æœ¬åˆ†æå·¥å…·"""
    print(f"   åŸå§‹æ–‡æœ¬: '{text}'")
    
    # åŸºç¡€ç»Ÿè®¡
    words = text.split()
    sentences = [s.strip() for s in text.split('.') if s.strip()]
    
    # è¯æ±‡åˆ†æ
    word_lengths = [len(word) for word in words]
    long_words = [word for word in words if len(word) > 5]
    
    # å­—ç¬¦åˆ†æ
    vowels = [char for char in text.lower() if char in 'aeiou']
    consonants = [char for char in text.lower() if char.isalpha() and char not in 'aeiou']
    
    # å¤§å†™å­—æ¯
    uppercase_chars = [char for char in text if char.isupper()]
    
    # æ•°å­—æå–
    numbers = [int(word) for word in words if word.isdigit()]
    
    # è¯é¢‘ç»Ÿè®¡
    word_freq = {word: words.count(word) for word in set(words)}
    common_words = [word for word, count in word_freq.items() if count > 1]
    
    print(f"\n   åˆ†æç»“æœ:")
    print(f"     å•è¯æ•°: {len(words)}")
    print(f"     å¥å­æ•°: {len(sentences)}")
    print(f"     å¹³å‡è¯é•¿: {sum(word_lengths) / len(word_lengths):.2f}")
    print(f"     é•¿å•è¯(>5å­—ç¬¦): {long_words}")
    print(f"     å…ƒéŸ³å­—æ¯æ•°: {len(vowels)}")
    print(f"     è¾…éŸ³å­—æ¯æ•°: {len(consonants)}")
    print(f"     å¤§å†™å­—æ¯: {uppercase_chars}")
    print(f"     æ•°å­—: {numbers}")
    print(f"     é‡å¤å•è¯: {common_words}")
    
    return {
        'word_count': len(words),
        'sentence_count': len(sentences),
        'avg_word_length': sum(word_lengths) / len(word_lengths),
        'long_words': long_words,
        'vowel_count': len(vowels),
        'consonant_count': len(consonants),
        'uppercase_chars': uppercase_chars,
        'numbers': numbers,
        'common_words': common_words
    }

# æµ‹è¯•æ–‡æœ¬åˆ†æ
sample_text = "Python is a powerful programming language. It has 3 major versions: Python 2, Python 3, and Python 4. Many developers love Python because it is easy to learn and use."
text_result = text_analysis(sample_text)

# æ¡ˆä¾‹3ï¼šæ•°å­¦è®¡ç®—å’Œç®—æ³•
print(f"\nğŸ“ 3. æ•°å­¦è®¡ç®—å’Œç®—æ³•:")

# æ–æ³¢é‚£å¥‘æ•°åˆ—
def fibonacci_sequence(n):
    """ç”Ÿæˆæ–æ³¢é‚£å¥‘æ•°åˆ—"""
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    else:
        fib = [0, 1]
        [fib.append(fib[-1] + fib[-2]) for _ in range(2, n)]
        return fib

# è´¨æ•°ç­›é€‰
def sieve_of_eratosthenes(limit):
    """åŸƒæ‹‰æ‰˜æ–¯ç‰¹å°¼ç­›æ³•"""
    if limit < 2:
        return []
    
    # åˆå§‹åŒ–ç­›å­
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False
    
    # ç­›é€‰è´¨æ•°
    [sieve.__setitem__(i, False) for i in range(2, int(limit ** 0.5) + 1) if sieve[i] 
     for j in range(i*i, limit + 1, i)]
    
    # æå–è´¨æ•°
    primes = [i for i, is_prime in enumerate(sieve) if is_prime]
    return primes

# å®Œç¾æ•°æŸ¥æ‰¾
def find_perfect_numbers(limit):
    """æŸ¥æ‰¾å®Œç¾æ•°"""
    def is_perfect(n):
        if n <= 1:
            return False
        divisors = [d for d in range(1, n) if n % d == 0]
        return sum(divisors) == n
    
    return [n for n in range(2, limit + 1) if is_perfect(n)]

# æµ‹è¯•æ•°å­¦ç®—æ³•
fib_numbers = fibonacci_sequence(10)
print(f"   æ–æ³¢é‚£å¥‘æ•°åˆ—(å‰10é¡¹): {fib_numbers}")

primes_up_to_30 = sieve_of_eratosthenes(30)
print(f"   30ä»¥å†…çš„è´¨æ•°: {primes_up_to_30}")

perfect_numbers = find_perfect_numbers(1000)
print(f"   1000ä»¥å†…çš„å®Œç¾æ•°: {perfect_numbers}")

# æ¡ˆä¾‹4ï¼šé…ç½®ç®¡ç†ç³»ç»Ÿ
print(f"\nğŸ“ 4. é…ç½®ç®¡ç†ç³»ç»Ÿ:")

class ConfigManager:
    def __init__(self):
        self.configs = {}
    
    def load_from_dict(self, config_dict):
        """ä»å­—å…¸åŠ è½½é…ç½®"""
        self.configs.update(config_dict)
    
    def load_from_list(self, config_list):
        """ä»åˆ—è¡¨åŠ è½½é…ç½®"""
        # æ ¼å¼: [(key, value), (key, value), ...]
        self.configs.update(dict(config_list))
    
    def get_filtered_configs(self, filter_func=None):
        """è·å–è¿‡æ»¤åçš„é…ç½®"""
        if filter_func:
            return {k: v for k, v in self.configs.items() if filter_func(k, v)}
        return self.configs.copy()
    
    def get_configs_by_type(self, value_type):
        """æŒ‰å€¼ç±»å‹è·å–é…ç½®"""
        return {k: v for k, v in self.configs.items() if isinstance(v, value_type)}
    
    def get_numeric_configs(self):
        """è·å–æ•°å€¼é…ç½®"""
        return {k: v for k, v in self.configs.items() if isinstance(v, (int, float))}
    
    def normalize_configs(self):
        """æ ‡å‡†åŒ–é…ç½®å€¼"""
        numeric_configs = self.get_numeric_configs()
        if numeric_configs:
            min_val = min(numeric_configs.values())
            max_val = max(numeric_configs.values())
            
            if max_val != min_val:
                return {
                    k: (v - min_val) / (max_val - min_val) 
                    for k, v in numeric_configs.items()
                }
        return {}

# æµ‹è¯•é…ç½®ç®¡ç†
config_manager = ConfigManager()

# åŠ è½½é…ç½®
config_data = {
    "database_url": "localhost:5432",
    "max_connections": 100,
    "timeout": 30.5,
    "debug_mode": True,
    "cache_size": 256,
    "log_level": "INFO",
    "retry_count": 3
}

config_manager.load_from_dict(config_data)
print(f"   æ‰€æœ‰é…ç½®: {config_manager.configs}")

# æŒ‰ç±»å‹è¿‡æ»¤
numeric_configs = config_manager.get_numeric_configs()
print(f"\n   æ•°å€¼é…ç½®: {numeric_configs}")

string_configs = config_manager.get_configs_by_type(str)
print(f"   å­—ç¬¦ä¸²é…ç½®: {string_configs}")

# è‡ªå®šä¹‰è¿‡æ»¤
large_configs = config_manager.get_filtered_configs(lambda k, v: isinstance(v, (int, float)) and v > 50)
print(f"\n   å¤§æ•°å€¼é…ç½®(>50): {large_configs}")

# æ ‡å‡†åŒ–é…ç½®
normalized = config_manager.normalize_configs()
print(f"\n   æ ‡å‡†åŒ–åçš„æ•°å€¼é…ç½®:")
for k, v in normalized.items():
    print(f"     {k}: {v:.3f}")

# æ¡ˆä¾‹5ï¼šå›¾åƒå¤„ç†æ¨¡æ‹Ÿ
print(f"\nğŸ“ 5. å›¾åƒå¤„ç†æ¨¡æ‹Ÿ:")

def simulate_image_processing(image_matrix):
    """æ¨¡æ‹Ÿå›¾åƒå¤„ç†"""
    print(f"   åŸå§‹å›¾åƒçŸ©é˜µ: {image_matrix}")
    
    # å›¾åƒåŸºæœ¬ä¿¡æ¯
    height = len(image_matrix)
    width = len(image_matrix[0]) if height > 0 else 0
    total_pixels = height * width
    
    print(f"   å›¾åƒå°ºå¯¸: {width}x{height}, æ€»åƒç´ : {total_pixels}")
    
    # æ‰å¹³åŒ–åƒç´ å€¼
    flat_pixels = [pixel for row in image_matrix for pixel in row]
    
    # äº®åº¦ç»Ÿè®¡
    min_brightness = min(flat_pixels)
    max_brightness = max(flat_pixels)
    avg_brightness = sum(flat_pixels) / len(flat_pixels)
    
    print(f"\n   äº®åº¦ç»Ÿè®¡:")
    print(f"     æœ€å°äº®åº¦: {min_brightness}")
    print(f"     æœ€å¤§äº®åº¦: {max_brightness}")
    print(f"     å¹³å‡äº®åº¦: {avg_brightness:.2f}")
    
    # äº®åº¦è°ƒæ•´
    # çº¿æ€§è°ƒæ•´
    adjusted_linear = [
        [int(pixel * 1.5) for pixel in row] 
        for row in image_matrix
    ]
    
    # é™åˆ¶äº®åº¦èŒƒå›´
    clamped = [
        [min(255, max(0, pixel)) for pixel in row]
        for row in adjusted_linear
    ]
    
    print(f"\n   çº¿æ€§äº®åº¦è°ƒæ•´å: {clamped}")
    
    # å¯¹æ¯”åº¦å¢å¼º
    contrast_factor = 1.5
    enhanced = [
        [int((pixel - avg_brightness) * contrast_factor + avg_brightness) 
         for pixel in row]
        for row in image_matrix
    ]
    
    # é™åˆ¶èŒƒå›´
    enhanced_clamped = [
        [min(255, max(0, pixel)) for pixel in row]
        for row in enhanced
    ]
    
    print(f"\n   å¯¹æ¯”åº¦å¢å¼ºå: {enhanced_clamped}")
    
    # è¾¹ç¼˜æ£€æµ‹æ¨¡æ‹Ÿ
    edges = [
        [
            abs(image_matrix[i][j] - image_matrix[i][j+1]) 
            if j+1 < width else 0
            for j in range(width)
        ]
        for i in range(height)
    ]
    
    print(f"\n   è¾¹ç¼˜æ£€æµ‹ç»“æœ: {edges}")
    
    return {
        'original': image_matrix,
        'adjusted': clamped,
        'enhanced': enhanced_clamped,
        'edges': edges,
        'stats': {
            'min_brightness': min_brightness,
            'max_brightness': max_brightness,
            'avg_brightness': avg_brightness
        }
    }

# æµ‹è¯•å›¾åƒå¤„ç†
test_image = [
    [50, 100, 150, 200],
    [75, 125, 175, 225],
    [100, 150, 200, 250],
    [25, 75, 125, 175]
]

image_result = simulate_image_processing(test_image)

print(f"\nâœ… æ¨å¯¼å¼å®é™…åº”ç”¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡æ¨å¯¼å¼å®é™…åº”ç”¨")
print(f"   âœ“ ç†è§£æ¨å¯¼å¼æ€§èƒ½ä¼˜åŒ–")
print(f"   âœ“ èƒ½ç‹¬ç«‹è®¾è®¡æ¨å¯¼å¼è§£å†³æ–¹æ¡ˆ")
````
