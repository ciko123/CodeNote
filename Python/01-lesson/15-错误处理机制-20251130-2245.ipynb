````python
print("ğŸ¨ è‡ªå®šä¹‰å¼‚å¸¸:")
print("=" * 50)

# åŸºæœ¬è‡ªå®šä¹‰å¼‚å¸¸
print(f"ğŸ“ 1. åŸºæœ¬è‡ªå®šä¹‰å¼‚å¸¸:")

class CustomError(Exception):
    """åŸºæœ¬çš„è‡ªå®šä¹‰å¼‚å¸¸"""
    def __init__(self, message, error_code=None):
        super().__init__(message)
        self.error_code = error_code
        self.timestamp = time.time()
    
    def __str__(self):
        if self.error_code:
            return f"[é”™è¯¯ç :{self.error_code}] {super().__str__()}"
        return super().__str__()

# æµ‹è¯•åŸºæœ¬è‡ªå®šä¹‰å¼‚å¸¸
try:
    raise CustomError("è¿™æ˜¯ä¸€ä¸ªè‡ªå®šä¹‰é”™è¯¯", error_code="E001")
except CustomError as e:
    print(f"   æ•è·è‡ªå®šä¹‰å¼‚å¸¸: {e}")
    print(f"   é”™è¯¯ç : {e.error_code}")
    print(f"   æ—¶é—´æˆ³: {e.timestamp}")
    print(f"   å¼‚å¸¸ç±»å‹: {type(e)}")

# é¢†åŸŸç‰¹å®šå¼‚å¸¸ä½“ç³»
print(f"\nğŸ“ 2. é¢†åŸŸç‰¹å®šå¼‚å¸¸ä½“ç³»:")

# é“¶è¡Œä¸šåŠ¡å¼‚å¸¸ä½“ç³»
class BankingError(Exception):
    """é“¶è¡Œä¸šåŠ¡åŸºç¡€å¼‚å¸¸"""
    def __init__(self, message, account_number=None):
        super().__init__(message)
        self.account_number = account_number
        self.timestamp = time.time()

class InsufficientFundsError(BankingError):
    """ä½™é¢ä¸è¶³å¼‚å¸¸"""
    def __init__(self, account_number, balance, amount):
        message = f"è´¦æˆ·{account_number}ä½™é¢ä¸è¶³: ä½™é¢{balance}, å°è¯•æå–{amount}"
        super().__init__(message, account_number)
        self.balance = balance
        self.amount = amount

class AccountNotFoundError(BankingError):
    """è´¦æˆ·ä¸å­˜åœ¨å¼‚å¸¸"""
    def __init__(self, account_number):
        message = f"è´¦æˆ·{account_number}ä¸å­˜åœ¨"
        super().__init__(message, account_number)

class InvalidTransactionError(BankingError):
    """æ— æ•ˆäº¤æ˜“å¼‚å¸¸"""
    def __init__(self, account_number, transaction_type, reason):
        message = f"è´¦æˆ·{account_number}çš„{transaction_type}äº¤æ˜“æ— æ•ˆ: {reason}"
        super().__init__(message, account_number)
        self.transaction_type = transaction_type
        self.reason = reason

# é“¶è¡Œè´¦æˆ·ç±»
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance
    
    def withdraw(self, amount):
        if amount <= 0:
            raise InvalidTransactionError(
                self.account_number, "å–æ¬¾", "é‡‘é¢å¿…é¡»å¤§äº0"
            )
        if amount > self.balance:
            raise InsufficientFundsError(
                self.account_number, self.balance, amount
            )
        
        self.balance -= amount
        return self.balance
    
    def deposit(self, amount):
        if amount <= 0:
            raise InvalidTransactionError(
                self.account_number, "å­˜æ¬¾", "é‡‘é¢å¿…é¡»å¤§äº0"
            )
        
        self.balance += amount
        return self.balance

# æµ‹è¯•é“¶è¡Œä¸šåŠ¡å¼‚å¸¸
print(f"   é“¶è¡Œä¸šåŠ¡å¼‚å¸¸æµ‹è¯•:")

account = BankAccount("12345", 1000)

# æ­£å¸¸æ“ä½œ
try:
    new_balance = account.withdraw(500)
    print(f"   å–æ¬¾æˆåŠŸï¼Œæ–°ä½™é¢: {new_balance}")
except BankingError as e:
    print(f"   é“¶è¡Œé”™è¯¯: {e}")

# ä½™é¢ä¸è¶³
try:
    account.withdraw(600)  # ä½™é¢åªæœ‰500
except InsufficientFundsError as e:
    print(f"   ä½™é¢ä¸è¶³é”™è¯¯: {e}")
    print(f"   å½“å‰ä½™é¢: {e.balance}, å°è¯•é‡‘é¢: {e.amount}")
except BankingError as e:
    print(f"   å…¶ä»–é“¶è¡Œé”™è¯¯: {e}")

# æ— æ•ˆäº¤æ˜“
try:
    account.deposit(-100)
except InvalidTransactionError as e:
    print(f"   æ— æ•ˆäº¤æ˜“é”™è¯¯: {e}")
    print(f"   äº¤æ˜“ç±»å‹: {e.transaction_type}, åŸå› : {e.reason}")
except BankingError as e:
    print(f"   å…¶ä»–é“¶è¡Œé”™è¯¯: {e}")

# æ•°æ®å¤„ç†å¼‚å¸¸ä½“ç³»
print(f"\nğŸ“ 3. æ•°æ®å¤„ç†å¼‚å¸¸ä½“ç³»:")

class DataProcessingError(Exception):
    """æ•°æ®å¤„ç†åŸºç¡€å¼‚å¸¸"""
    def __init__(self, message, data_source=None, row_index=None):
        super().__init__(message)
        self.data_source = data_source
        self.row_index = row_index
        self.timestamp = time.time()

class DataValidationError(DataProcessingError):
    """æ•°æ®éªŒè¯å¼‚å¸¸"""
    def __init__(self, data_source, row_index, field, value, validation_rule):
        message = f"æ•°æ®æº{data_source}ç¬¬{row_index}è¡Œå­—æ®µ{field}å€¼{value}ä¸ç¬¦åˆè§„åˆ™: {validation_rule}"
        super().__init__(message, data_source, row_index)
        self.field = field
        self.value = value
        self.validation_rule = validation_rule

class DataFormatError(DataProcessingError):
    """æ•°æ®æ ¼å¼å¼‚å¸¸"""
    def __init__(self, data_source, expected_format, actual_format):
        message = f"æ•°æ®æº{data_source}æ ¼å¼é”™è¯¯: æœŸæœ›{expected_format}, å®é™…{actual_format}"
        super().__init__(message, data_source)
        self.expected_format = expected_format
        self.actual_format = actual_format

class DataTransformationError(DataProcessingError):
    """æ•°æ®è½¬æ¢å¼‚å¸¸"""
    def __init__(self, data_source, row_index, transformation, error_detail):
        message = f"æ•°æ®æº{data_source}ç¬¬{row_index}è¡Œè½¬æ¢{transformation}å¤±è´¥: {error_detail}"
        super().__init__(message, data_source, row_index)
        self.transformation = transformation
        self.error_detail = error_detail

# æ•°æ®å¤„ç†å™¨ç±»
class DataProcessor:
    def __init__(self, data_source):
        self.data_source = data_source
    
    def validate_row(self, row_data, row_index):
        """éªŒè¯è¡Œæ•°æ®"""
        required_fields = ['name', 'age', 'email']
        
        for field in required_fields:
            if field not in row_data:
                raise DataValidationError(
                    self.data_source, row_index, field, None, f"ç¼ºå°‘å¿…éœ€å­—æ®µ{field}"
                )
        
        # å¹´é¾„éªŒè¯
        age = row_data.get('age')
        if not isinstance(age, int) or age < 0 or age > 150:
            raise DataValidationError(
                self.data_source, row_index, 'age', age, "å¹´é¾„å¿…é¡»æ˜¯0-150ä¹‹é—´çš„æ•´æ•°"
            )
        
        # é‚®ç®±éªŒè¯
        email = row_data.get('email', '')
        if '@' not in email:
            raise DataValidationError(
                self.data_source, row_index, 'email', email, "é‚®ç®±å¿…é¡»åŒ…å«@ç¬¦å·"
            )
    
    def transform_row(self, row_data, row_index):
        """è½¬æ¢è¡Œæ•°æ®"""
        try:
            # è½¬æ¢å¹´é¾„ä¸ºå­—ç¬¦ä¸²
            transformed = row_data.copy()
            transformed['age_str'] = str(row_data['age'])
            
            # è½¬æ¢é‚®ç®±ä¸ºå°å†™
            transformed['email_lower'] = row_data['email'].lower()
            
            return transformed
        except Exception as e:
            raise DataTransformationError(
                self.data_source, row_index, "æ•°æ®è½¬æ¢", str(e)
            ) from e

# æµ‹è¯•æ•°æ®å¤„ç†å¼‚å¸¸
print(f"   æ•°æ®å¤„ç†å¼‚å¸¸æµ‹è¯•:")

processor = DataProcessor("ç”¨æˆ·æ•°æ®è¡¨")

# æµ‹è¯•æ•°æ®
test_rows = [
    {'name': 'å¼ ä¸‰', 'age': 25, 'email': 'zhangsan@example.com'},  # æ­£å¸¸æ•°æ®
    {'name': 'æå››', 'age': -5, 'email': 'lisi@example.com'},  # å¹´é¾„æ— æ•ˆ
    {'name': 'ç‹äº”', 'age': 30, 'email': 'invalid_email'},  # é‚®ç®±æ— æ•ˆ
    {'name': 'èµµå…­', 'age': 'abc', 'email': 'zhaoliu@example.com'},  # å¹´é¾„ç±»å‹é”™è¯¯
]

for i, row in enumerate(test_rows, 1):
    try:
        processor.validate_row(row, i)
        transformed = processor.transform_row(row, i)
        print(f"   ç¬¬{i}è¡Œå¤„ç†æˆåŠŸ: {transformed['name']}")
    except DataValidationError as e:
        print(f"   ç¬¬{i}è¡ŒéªŒè¯å¤±è´¥: {e}")
    except DataTransformationError as e:
        print(f"   ç¬¬{i}è¡Œè½¬æ¢å¤±è´¥: {e}")
    except DataProcessingError as e:
        print(f"   ç¬¬{i}è¡Œå¤„ç†é”™è¯¯: {e}")

# å¼‚å¸¸é“¾å’Œä¸Šä¸‹æ–‡
print(f"\nğŸ“ 4. å¼‚å¸¸é“¾å’Œä¸Šä¸‹æ–‡:")

class DatabaseError(Exception):
    """æ•°æ®åº“é”™è¯¯"""
    def __init__(self, message, query=None, params=None):
        super().__init__(message)
        self.query = query
        self.params = params

class UserService:
    def __init__(self, db_connection):
        self.db_connection = db_connection
    
    def get_user(self, user_id):
        """è·å–ç”¨æˆ·ä¿¡æ¯"""
        try:
            # æ¨¡æ‹Ÿæ•°æ®åº“æŸ¥è¯¢
            if user_id == 999:
                raise ConnectionError("æ•°æ®åº“è¿æ¥å¤±è´¥")
            elif user_id == 888:
                raise ValueError("æ— æ•ˆçš„ç”¨æˆ·IDæ ¼å¼")
            else:
                return {"id": user_id, "name": f"ç”¨æˆ·{user_id}"}
        
        except ConnectionError as e:
            raise DatabaseError(
                f"è·å–ç”¨æˆ·{user_id}æ—¶æ•°æ®åº“è¿æ¥å¤±è´¥",
                query="SELECT * FROM users WHERE id = ?",
                params=[user_id]
            ) from e
        
        except ValueError as e:
            raise DatabaseError(
                f"è·å–ç”¨æˆ·{user_id}æ—¶å‚æ•°é”™è¯¯",
                query="SELECT * FROM users WHERE id = ?",
                params=[user_id]
            ) from e

# æµ‹è¯•å¼‚å¸¸é“¾
print(f"   å¼‚å¸¸é“¾æµ‹è¯•:")

user_service = UserService("æ¨¡æ‹Ÿæ•°æ®åº“è¿æ¥")

test_user_ids = [123, 999, 888]

for user_id in test_user_ids:
    try:
        user = user_service.get_user(user_id)
        print(f"   è·å–ç”¨æˆ·{user_id}æˆåŠŸ: {user}")
    except DatabaseError as e:
        print(f"   æ•°æ®åº“é”™è¯¯: {e}")
        print(f"   æŸ¥è¯¢: {e.query}")
        print(f"   å‚æ•°: {e.params}")
        print(f"   åŸå§‹å¼‚å¸¸: {type(e.__cause__).__name__}: {e.__cause__}")

print(f"\nâœ… è‡ªå®šä¹‰å¼‚å¸¸å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡è‡ªå®šä¹‰å¼‚å¸¸å®šä¹‰æ–¹æ³•")
print(f"   âœ“ ç†è§£å¼‚å¸¸ç»§æ‰¿å’Œæ‰©å±•")
print(f"   âœ“ èƒ½è®¾è®¡é¢†åŸŸç‰¹å®šå¼‚å¸¸ä½“ç³»")
````
