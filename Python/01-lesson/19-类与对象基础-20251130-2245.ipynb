````python
print("ğŸ—ï¸ å®é™…åº”ç”¨æ¡ˆä¾‹:")
print("=" * 50)

# æ¡ˆä¾‹1: å›¾ä¹¦ç®¡ç†ç³»ç»Ÿ
print(f"ğŸ“ 1. å›¾ä¹¦ç®¡ç†ç³»ç»Ÿ:")

class Book:
    """å›¾ä¹¦ç±»"""
    
    def __init__(self, isbn, title, author, publisher, year):
        self.isbn = isbn
        self.title = title
        self.author = author
        self.publisher = publisher
        self.year = year
        self.is_borrowed = False
        self.borrower = None
        self.borrow_date = None
    
    def borrow(self, borrower_name, borrow_date):
        """å€Ÿä¹¦"""
        if not self.is_borrowed:
            self.is_borrowed = True
            self.borrower = borrower_name
            self.borrow_date = borrow_date
            print(f"   ã€Š{self.title}ã€‹å·²å€Ÿç»™ {borrower_name}")
            return True
        else:
            print(f"   ã€Š{self.title}ã€‹å·²è¢« {self.borrower} å€Ÿå‡º")
            return False
    
    def return_book(self):
        """è¿˜ä¹¦"""
        if self.is_borrowed:
            print(f"   ã€Š{self.title}ã€‹å·²å½’è¿˜")
            self.is_borrowed = False
            self.borrower = None
            self.borrow_date = None
            return True
        else:
            print(f"   ã€Š{self.title}ã€‹æœªè¢«å€Ÿå‡º")
            return False
    
    def get_info(self):
        """è·å–å›¾ä¹¦ä¿¡æ¯"""
        status = "å·²å€Ÿå‡º" if self.is_borrowed else "å¯å€Ÿé˜…"
        info = f"ISBN: {self.isbn}, ã€Š{self.title}ã€‹, {self.author}, {status}"
        if self.is_borrowed:
            info += f", å€Ÿé˜…è€…: {self.borrower}"
        return info

class Library:
    """å›¾ä¹¦é¦†ç±»"""
    
    def __init__(self, name):
        self.name = name
        self.books = {}  # ISBN -> Book
        self.borrowed_books = []
    
    def add_book(self, book):
        """æ·»åŠ å›¾ä¹¦"""
        if book.isbn not in self.books:
            self.books[book.isbn] = book
            print(f"   å›¾ä¹¦ã€Š{book.title}ã€‹å·²æ·»åŠ åˆ° {self.name}")
        else:
            print(f"   ISBN {book.isbn} çš„å›¾ä¹¦å·²å­˜åœ¨")
    
    def find_book(self, isbn):
        """æŸ¥æ‰¾å›¾ä¹¦"""
        return self.books.get(isbn)
    
    def search_books(self, keyword):
        """æœç´¢å›¾ä¹¦"""
        results = []
        for book in self.books.values():
            if (keyword.lower() in book.title.lower() or 
                keyword.lower() in book.author.lower()):
                results.append(book)
        return results
    
    def list_available_books(self):
        """åˆ—å‡ºå¯å€Ÿé˜…å›¾ä¹¦"""
        available = [book for book in self.books.values() if not book.is_borrowed]
        return available
    
    def list_borrowed_books(self):
        """åˆ—å‡ºå·²å€Ÿå‡ºå›¾ä¹¦"""
        borrowed = [book for book in self.books.values() if book.is_borrowed]
        return borrowed
    
    def get_statistics(self):
        """è·å–ç»Ÿè®¡ä¿¡æ¯"""
        total = len(self.books)
        borrowed = len(self.list_borrowed_books())
        available = total - borrowed
        return {
            "total_books": total,
            "borrowed_books": borrowed,
            "available_books": available
        }

# åˆ›å»ºå›¾ä¹¦é¦†å’Œå›¾ä¹¦
library = Library("å¸‚å›¾ä¹¦é¦†")

# æ·»åŠ å›¾ä¹¦
books = [
    Book("978-7-111-12345-6", "Pythonç¼–ç¨‹", "å¼ ä¸‰", "æœºæ¢°å·¥ä¸šå‡ºç‰ˆç¤¾", 2020),
    Book("978-7-115-23456-7", "æ•°æ®ç»“æ„", "æå››", "äººæ°‘é‚®ç”µå‡ºç‰ˆç¤¾", 2019),
    Book("978-7-302-34567-8", "ç®—æ³•å¯¼è®º", "ç‹äº”", "æ¸…åå¤§å­¦å‡ºç‰ˆç¤¾", 2018),
    Book("978-7-121-45678-9", "æœºå™¨å­¦ä¹ ", "èµµå…­", "ç”µå­å·¥ä¸šå‡ºç‰ˆç¤¾", 2021)
]

for book in books:
    library.add_book(book)

print(f"\n   å›¾ä¹¦é¦†ç»Ÿè®¡: {library.get_statistics()}")

# å€Ÿä¹¦æ“ä½œ
print(f"\n   å€Ÿä¹¦æ“ä½œ:")
book1 = library.find_book("978-7-111-12345-6")
book2 = library.find_book("978-7-115-23456-7")

book1.borrow("å°æ˜", "2023-12-01")
book2.borrow("å°çº¢", "2023-12-02")

print(f"\n   å€Ÿä¹¦åç»Ÿè®¡: {library.get_statistics()}")

# æœç´¢å›¾ä¹¦
print(f"\n   æœç´¢åŒ…å«'Python'çš„å›¾ä¹¦:")
search_results = library.search_books("Python")
for book in search_results:
    print(f"     {book.get_info()}")

# æ¡ˆä¾‹2: é“¶è¡Œè´¦æˆ·ç³»ç»Ÿ
print(f"\nğŸ“ 2. é“¶è¡Œè´¦æˆ·ç³»ç»Ÿ:")

class BankAccount:
    """é“¶è¡Œè´¦æˆ·ç±»"""
    
    # ç±»å±æ€§
    bank_name = "ç¤ºä¾‹é“¶è¡Œ"
    interest_rate = 0.03  # å¹´åˆ©ç‡
    
    def __init__(self, account_number, account_holder, initial_balance=0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = float(initial_balance)
        self.transaction_history = []
        self.is_active = True
        self.open_date = "2023-01-01"
    
    def deposit(self, amount):
        """å­˜æ¬¾"""
        if amount > 0:
            self.balance += amount
            self.transaction_history.append(f"å­˜æ¬¾: +{amount:.2f}")
            print(f"   {self.account_holder} å­˜æ¬¾ {amount:.2f} æˆåŠŸ")
            return True
        else:
            print(f"   å­˜æ¬¾é‡‘é¢å¿…é¡»å¤§äº0")
            return False
    
    def withdraw(self, amount):
        """å–æ¬¾"""
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            self.transaction_history.append(f"å–æ¬¾: -{amount:.2f}")
            print(f"   {self.account_holder} å–æ¬¾ {amount:.2f} æˆåŠŸ")
            return True
        else:
            print(f"   å–æ¬¾å¤±è´¥ï¼šä½™é¢ä¸è¶³æˆ–é‡‘é¢æ— æ•ˆ")
            return False
    
    def transfer(self, target_account, amount):
        """è½¬è´¦"""
        if self.withdraw(amount):
            target_account.deposit(amount)
            self.transaction_history.append(f"è½¬è´¦ç»™{target_account.account_holder}: -{amount:.2f}")
            target_account.transaction_history.append(f"æ”¶åˆ°{self.account_holder}è½¬è´¦: +{amount:.2f}")
            print(f"   å‘ {target_account.account_holder} è½¬è´¦ {amount:.2f} æˆåŠŸ")
            return True
        return False
    
    def calculate_interest(self, months):
        """è®¡ç®—åˆ©æ¯"""
        interest = self.balance * (BankAccount.interest_rate / 12) * months
        return interest
    
    def get_account_info(self):
        """è·å–è´¦æˆ·ä¿¡æ¯"""
        status = "æ´»è·ƒ" if self.is_active else "å†»ç»“"
        return f"{self.account_number}: {self.account_holder}, ä½™é¢: {self.balance:.2f}, çŠ¶æ€: {status}"

class SavingsAccount(BankAccount):
    """å‚¨è“„è´¦æˆ·ç±»"""
    
    def __init__(self, account_number, account_holder, initial_balance=0):
        super().__init__(account_number, account_holder, initial_balance)
        self.account_type = "å‚¨è“„è´¦æˆ·"
        self.withdraw_limit = 50000  # å–æ¬¾é™é¢
    
    def withdraw(self, amount):
        """é‡å†™å–æ¬¾æ–¹æ³•ï¼Œå¢åŠ é™é¢æ£€æŸ¥"""
        if amount > self.withdraw_limit:
            print(f"   å–æ¬¾é‡‘é¢è¶…è¿‡é™é¢ {self.withdraw_limit}")
            return False
        return super().withdraw(amount)

class CheckingAccount(BankAccount):
    """æ”¯ç¥¨è´¦æˆ·ç±»"""
    
    def __init__(self, account_number, account_holder, initial_balance=0):
        super().__init__(account_number, account_holder, initial_balance)
        self.account_type = "æ”¯ç¥¨è´¦æˆ·"
        self.overdraft_limit = 10000  # é€æ”¯é™é¢
    
    def withdraw(self, amount):
        """é‡å†™å–æ¬¾æ–¹æ³•ï¼Œæ”¯æŒé€æ”¯"""
        if amount > 0 and amount <= (self.balance + self.overdraft_limit):
            self.balance -= amount
            self.transaction_history.append(f"å–æ¬¾: -{amount:.2f}")
            print(f"   {self.account_holder} å–æ¬¾ {amount:.2f} æˆåŠŸï¼ˆå¯èƒ½é€æ”¯ï¼‰")
            return True
        else:
            print(f"   å–æ¬¾å¤±è´¥ï¼šè¶…è¿‡é€æ”¯é™é¢")
            return False

# åˆ›å»ºé“¶è¡Œè´¦æˆ·
print(f"   åˆ›å»ºé“¶è¡Œè´¦æˆ·:")
savings_acc = SavingsAccount("SA001", "å¼ ä¸‰", 10000)
checking_acc = CheckingAccount("CA001", "æå››", 5000)

print(f"\n   è´¦æˆ·ä¿¡æ¯:")
print(f"     {savings_acc.get_account_info()}")
print(f"     {checking_acc.get_account_info()}")

# è¿›è¡Œäº¤æ˜“
print(f"\n   è¿›è¡Œäº¤æ˜“:")
savings_acc.deposit(2000)
checking_acc.withdraw(8000)  # é€æ”¯
savings_acc.transfer(checking_acc, 3000)

print(f"\n   äº¤æ˜“åè´¦æˆ·ä¿¡æ¯:")
print(f"     {savings_acc.get_account_info()}")
print(f"     {checking_acc.get_account_info()}")

# æ¡ˆä¾‹3: ç”µå•†è®¢å•ç³»ç»Ÿ
print(f"\nğŸ“ 3. ç”µå•†è®¢å•ç³»ç»Ÿ:")

class Product:
    """å•†å“ç±»"""
    
    def __init__(self, product_id, name, price, stock):
        self.product_id = product_id
        self.name = name
        self.price = float(price)
        self.stock = int(stock)
        self.description = ""
    
    def update_stock(self, quantity):
        """æ›´æ–°åº“å­˜"""
        self.stock += quantity
        print(f"   å•†å“ {self.name} åº“å­˜æ›´æ–°ä¸º {self.stock}")
    
    def is_available(self, quantity):
        """æ£€æŸ¥æ˜¯å¦æœ‰è¶³å¤Ÿåº“å­˜"""
        return self.stock >= quantity

class OrderItem:
    """è®¢å•é¡¹ç±»"""
    
    def __init__(self, product, quantity):
        self.product = product
        self.quantity = quantity
        self.unit_price = product.price
        self.total_price = self.unit_price * self.quantity
    
    def get_item_info(self):
        """è·å–è®¢å•é¡¹ä¿¡æ¯"""
        return f"{self.product.name} x{self.quantity} = {self.total_price:.2f}"

class Order:
    """è®¢å•ç±»"""
    
    def __init__(self, order_id, customer_name):
        self.order_id = order_id
        self.customer_name = customer_name
        self.items = []
        self.order_date = "2023-12-01"
        self.status = "å¾…å¤„ç†"
        self.total_amount = 0.0
    
    def add_item(self, product, quantity):
        """æ·»åŠ è®¢å•é¡¹"""
        if product.is_available(quantity):
            item = OrderItem(product, quantity)
            self.items.append(item)
            self.total_amount += item.total_price
            print(f"   æ·»åŠ å•†å“: {item.get_item_info()}")
            return True
        else:
            print(f"   å•†å“ {product.name} åº“å­˜ä¸è¶³")
            return False
    
    def remove_item(self, product_id):
        """ç§»é™¤è®¢å•é¡¹"""
        for i, item in enumerate(self.items):
            if item.product.product_id == product_id:
                removed_item = self.items.pop(i)
                self.total_amount -= removed_item.total_price
                print(f"   ç§»é™¤å•†å“: {removed_item.get_item_info()}")
                return True
        print(f"   æœªæ‰¾åˆ°å•†å“ID {product_id}")
        return False
    
    def calculate_total(self):
        """è®¡ç®—è®¢å•æ€»é¢"""
        self.total_amount = sum(item.total_price for item in self.items)
        return self.total_amount
    
    def confirm_order(self):
        """ç¡®è®¤è®¢å•"""
        if self.items:
            self.status = "å·²ç¡®è®¤"
            # å‡å°‘å•†å“åº“å­˜
            for item in self.items:
                item.product.update_stock(-item.quantity)
            print(f"   è®¢å• {self.order_id} å·²ç¡®è®¤")
            return True
        else:
            print(f"   è®¢å•ä¸ºç©ºï¼Œæ— æ³•ç¡®è®¤")
            return False
    
    def get_order_summary(self):
        """è·å–è®¢å•æ‘˜è¦"""
        summary = f"è®¢å• {self.order_id} - {self.customer_name}\n"
        summary += f"çŠ¶æ€: {self.status}, æ€»é¢: {self.total_amount:.2f}\n"
        summary += "å•†å“åˆ—è¡¨:\n"
        for item in self.items:
            summary += f"  {item.get_item_info()}\n"
        return summary

# åˆ›å»ºå•†å“
products = [
    Product("P001", "ç¬”è®°æœ¬ç”µè„‘", 5999.99, 10),
    Product("P002", "æ— çº¿é¼ æ ‡", 99.99, 50),
    Product("P003", "æœºæ¢°é”®ç›˜", 299.99, 30),
    Product("P004", "æ˜¾ç¤ºå™¨", 1299.99, 15)
]

# åˆ›å»ºè®¢å•
order = Order("ORD001", "ç‹å…ˆç”Ÿ")

print(f"   åˆ›å»ºè®¢å•:")
order.add_item(products[0], 1)  # ç¬”è®°æœ¬ç”µè„‘
order.add_item(products[1], 2)  # æ— çº¿é¼ æ ‡
order.add_item(products[2], 1)  # æœºæ¢°é”®ç›˜

print(f"\n   è®¢å•æ‘˜è¦:")
print(order.get_order_summary())

# ç¡®è®¤è®¢å•
order.confirm_order()

print(f"\n   ç¡®è®¤åå•†å“åº“å­˜:")
for product in products:
    print(f"     {product.name}: {product.stock} ä»¶")

print(f"\nâœ… å®é™…åº”ç”¨æ¡ˆä¾‹å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡é¢å‘å¯¹è±¡è®¾è®¡æ€è·¯")
print(f"   âœ“ ç†è§£ç±»å’Œå¯¹è±¡åœ¨å®é™…é—®é¢˜ä¸­çš„ä½œç”¨")
print(f"   âœ“ èƒ½åº”ç”¨é¢å‘å¯¹è±¡è§£å†³å¤æ‚é—®é¢˜")
````
