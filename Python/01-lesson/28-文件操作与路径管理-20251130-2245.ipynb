````python
print("ğŸ“ æ–‡ä»¶åŸºæœ¬æ“ä½œ:")
print("=" * 50)

# 1. æ–‡ä»¶åŸºæœ¬è¯»å†™æ“ä½œ
print(f"ğŸ“ 1. æ–‡ä»¶åŸºæœ¬è¯»å†™æ“ä½œ:")

# åˆ›å»ºæµ‹è¯•ç›®å½•
import os
test_dir = "test_files"
if not os.path.exists(test_dir):
    os.makedirs(test_dir)
    print(f"   åˆ›å»ºæµ‹è¯•ç›®å½•: {test_dir}")

# 1.1 åŸºæœ¬å†™å…¥æ“ä½œ
print(f"\n   ğŸ“ 1.1 åŸºæœ¬å†™å…¥æ“ä½œ:")

# å†™å…¥æ–‡æœ¬æ–‡ä»¶
text_content = """Pythonæ–‡ä»¶æ“ä½œç¤ºä¾‹
è¿™æ˜¯ç¬¬ä¸€è¡Œ
è¿™æ˜¯ç¬¬äºŒè¡Œ
è¿™æ˜¯ç¬¬ä¸‰è¡Œ
"""

file_path = os.path.join(test_dir, "sample.txt")
with open(file_path, 'w', encoding='utf-8') as f:
    f.write(text_content)
    print(f"   æ–‡ä»¶å†™å…¥æˆåŠŸ: {file_path}")

# 1.2 åŸºæœ¬è¯»å–æ“ä½œ
print(f"\n   ğŸ“– 1.2 åŸºæœ¬è¯»å–æ“ä½œ:")

# è¯»å–æ•´ä¸ªæ–‡ä»¶
with open(file_path, 'r', encoding='utf-8') as f:
    content = f.read()
    print(f"   æ–‡ä»¶å†…å®¹:")
    print(f"   {repr(content)}")

# é€è¡Œè¯»å–
print(f"\n   ğŸ“„ é€è¡Œè¯»å–:")
with open(file_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()
    print(f"   æ–‡ä»¶è¡Œæ•°: {len(lines)}")
    for i, line in enumerate(lines, 1):
        print(f"     ç¬¬{i}è¡Œ: {repr(line)}")

# 1.3 è¿½åŠ å†™å…¥
print(f"\n   â• 1.3 è¿½åŠ å†™å…¥:")

additional_content = "è¿™æ˜¯è¿½åŠ çš„å†…å®¹\n"
with open(file_path, 'a', encoding='utf-8') as f:
    f.write(additional_content)
    print(f"   è¿½åŠ å†…å®¹æˆåŠŸ")

# éªŒè¯è¿½åŠ ç»“æœ
with open(file_path, 'r', encoding='utf-8') as f:
    final_content = f.read()
    print(f"   è¿½åŠ åæ–‡ä»¶å†…å®¹:")
    print(f"   {repr(final_content)}")

# 1.4 ä¸åŒå†™å…¥æ¨¡å¼
print(f"\n   ğŸ“ 1.4 ä¸åŒå†™å…¥æ¨¡å¼:")

# å†™å…¥åˆ—è¡¨å†…å®¹
lines_to_write = [
    "åˆ—è¡¨ç¬¬ä¸€è¡Œ\n",
    "åˆ—è¡¨ç¬¬äºŒè¡Œ\n", 
    "åˆ—è¡¨ç¬¬ä¸‰è¡Œ\n"
]

list_file = os.path.join(test_dir, "lines.txt")
with open(list_file, 'w', encoding='utf-8') as f:
    f.writelines(lines_to_write)
    print(f"   åˆ—è¡¨å†…å®¹å†™å…¥æˆåŠŸ: {list_file}")

# 2. æ–‡ä»¶ä¸Šä¸‹æ–‡ç®¡ç†å™¨
print(f"\nğŸ“ 2. æ–‡ä»¶ä¸Šä¸‹æ–‡ç®¡ç†å™¨:")
print(f"   ä½¿ç”¨withè¯­å¥è‡ªåŠ¨ç®¡ç†æ–‡ä»¶èµ„æº")

class FileHandler:
    """æ–‡ä»¶å¤„ç†å™¨ç±»"""
    
    def __init__(self, file_path, mode='r', encoding='utf-8'):
        self.file_path = file_path
        self.mode = mode
        self.encoding = encoding
    
    def read_file(self):
        """å®‰å…¨è¯»å–æ–‡ä»¶"""
        try:
            with open(self.file_path, self.mode, encoding=self.encoding) as f:
                return f.read()
        except FileNotFoundError:
            print(f"   é”™è¯¯: æ–‡ä»¶ä¸å­˜åœ¨ {self.file_path}")
            return None
        except UnicodeDecodeError:
            print(f"   é”™è¯¯: æ–‡ä»¶ç¼–ç é—®é¢˜ {self.file_path}")
            return None
        except Exception as e:
            print(f"   é”™è¯¯: è¯»å–æ–‡ä»¶æ—¶å‘ç”Ÿå¼‚å¸¸ {e}")
            return None
    
    def write_file(self, content):
        """å®‰å…¨å†™å…¥æ–‡ä»¶"""
        try:
            with open(self.file_path, 'w', encoding=self.encoding) as f:
                f.write(content)
            print(f"   æ–‡ä»¶å†™å…¥æˆåŠŸ: {self.file_path}")
            return True
        except Exception as e:
            print(f"   é”™è¯¯: å†™å…¥æ–‡ä»¶æ—¶å‘ç”Ÿå¼‚å¸¸ {e}")
            return False
    
    def append_file(self, content):
        """å®‰å…¨è¿½åŠ æ–‡ä»¶"""
        try:
            with open(self.file_path, 'a', encoding=self.encoding) as f:
                f.write(content)
            print(f"   æ–‡ä»¶è¿½åŠ æˆåŠŸ: {self.file_path}")
            return True
        except Exception as e:
            print(f"   é”™è¯¯: è¿½åŠ æ–‡ä»¶æ—¶å‘ç”Ÿå¼‚å¸¸ {e}")
            return False
    
    def copy_file(self, destination):
        """å¤åˆ¶æ–‡ä»¶"""
        try:
            import shutil
            shutil.copy2(self.file_path, destination)
            print(f"   æ–‡ä»¶å¤åˆ¶æˆåŠŸ: {self.file_path} -> {destination}")
            return True
        except Exception as e:
            print(f"   é”™è¯¯: å¤åˆ¶æ–‡ä»¶æ—¶å‘ç”Ÿå¼‚å¸¸ {e}")
            return False

# æµ‹è¯•æ–‡ä»¶å¤„ç†å™¨
print(f"\n   æµ‹è¯•æ–‡ä»¶å¤„ç†å™¨:")

handler = FileHandler(file_path)
content = handler.read_file()
if content:
    print(f"     è¯»å–å†…å®¹é•¿åº¦: {len(content)} å­—ç¬¦")

# æµ‹è¯•å†™å…¥æ–°æ–‡ä»¶
new_file = os.path.join(test_dir, "handler_test.txt")
new_handler = FileHandler(new_file, 'w')
new_handler.write_file("ä½¿ç”¨æ–‡ä»¶å¤„ç†å™¨å†™å…¥çš„å†…å®¹\nç¬¬äºŒè¡Œå†…å®¹\n")

# 3. æ–‡ä»¶ç¼–ç å¤„ç†
print(f"\nğŸ“ 3. æ–‡ä»¶ç¼–ç å¤„ç†:")

# 3.1 ä¸åŒç¼–ç å†™å…¥
print(f"\n   ğŸ“ 3.1 ä¸åŒç¼–ç å†™å…¥:")

# UTF-8ç¼–ç 
utf8_file = os.path.join(test_dir, "utf8.txt")
chinese_content = "è¿™æ˜¯ä¸­æ–‡å†…å®¹\nEnglish content\næ—¥æœ¬èªã®å†…å®¹\n"

with open(utf8_file, 'w', encoding='utf-8') as f:
    f.write(chinese_content)
    print(f"   UTF-8æ–‡ä»¶å†™å…¥æˆåŠŸ")

# GBKç¼–ç 
gbk_file = os.path.join(test_dir, "gbk.txt")
with open(gbk_file, 'w', encoding='gbk') as f:
    f.write(chinese_content)
    print(f"   GBKæ–‡ä»¶å†™å…¥æˆåŠŸ")

# 3.2 ç¼–ç æ£€æµ‹å’Œè½¬æ¢
print(f"\n   ğŸ” 3.2 ç¼–ç æ£€æµ‹å’Œè½¬æ¢:")

def detect_encoding(file_path):
    """æ£€æµ‹æ–‡ä»¶ç¼–ç """
    try:
        import chardet
        with open(file_path, 'rb') as f:
            raw_data = f.read()
            result = chardet.detect(raw_data)
            return result['encoding'], result['confidence']
    except ImportError:
        print(f"   chardetæ¨¡å—æœªå®‰è£…ï¼Œä½¿ç”¨é»˜è®¤æ£€æµ‹")
        return 'utf-8', 1.0
    except Exception as e:
        print(f"   ç¼–ç æ£€æµ‹å¤±è´¥: {e}")
        return None, 0

def convert_encoding(input_file, output_file, from_encoding, to_encoding):
    """è½¬æ¢æ–‡ä»¶ç¼–ç """
    try:
        with open(input_file, 'r', encoding=from_encoding) as f:
            content = f.read()
        
        with open(output_file, 'w', encoding=to_encoding) as f:
            f.write(content)
        
        print(f"   ç¼–ç è½¬æ¢æˆåŠŸ: {from_encoding} -> {to_encoding}")
        return True
    except Exception as e:
        print(f"   ç¼–ç è½¬æ¢å¤±è´¥: {e}")
        return False

# æ£€æµ‹ç¼–ç 
encoding, confidence = detect_encoding(utf8_file)
if encoding:
    print(f"   æ£€æµ‹åˆ°ç¼–ç : {encoding} (ç½®ä¿¡åº¦: {confidence:.2f})")

# è½¬æ¢ç¼–ç 
converted_file = os.path.join(test_dir, "converted.txt")
convert_encoding(utf8_file, converted_file, 'utf-8', 'gbk')

# 4. æ–‡ä»¶å±æ€§å’Œå…ƒæ•°æ®
print(f"\nğŸ“ 4. æ–‡ä»¶å±æ€§å’Œå…ƒæ•°æ®:")

def get_file_info(file_path):
    """è·å–æ–‡ä»¶ä¿¡æ¯"""
    try:
        stat = os.stat(file_path)
        return {
            'size': stat.st_size,
            'created': stat.st_ctime,
            'modified': stat.st_mtime,
            'accessed': stat.st_atime,
            'is_file': os.path.isfile(file_path),
            'is_dir': os.path.isdir(file_path),
            'exists': os.path.exists(file_path)
        }
    except Exception as e:
        print(f"   è·å–æ–‡ä»¶ä¿¡æ¯å¤±è´¥: {e}")
        return None

def format_file_info(info):
    """æ ¼å¼åŒ–æ–‡ä»¶ä¿¡æ¯"""
    if not info:
        return "æ— æ–‡ä»¶ä¿¡æ¯"
    
    import datetime
    
    created = datetime.datetime.fromtimestamp(info['created'])
    modified = datetime.datetime.fromtimestamp(info['modified'])
    accessed = datetime.datetime.fromtimestamp(info['accessed'])
    
    return f"""
    æ–‡ä»¶å¤§å°: {info['size']} å­—èŠ‚
    åˆ›å»ºæ—¶é—´: {created.strftime('%Y-%m-%d %H:%M:%S')}
    ä¿®æ”¹æ—¶é—´: {modified.strftime('%Y-%m-%d %H:%M:%S')}
    è®¿é—®æ—¶é—´: {accessed.strftime('%Y-%m-%d %H:%M:%S')}
    æ˜¯æ–‡ä»¶: {info['is_file']}
    æ˜¯ç›®å½•: {info['is_dir']}
    å­˜åœ¨: {info['exists']}
    """

# è·å–æ–‡ä»¶ä¿¡æ¯
file_info = get_file_info(file_path)
print(f"\n   æ–‡ä»¶ä¿¡æ¯:")
print(format_file_info(file_info))

print(f"\nâœ… æ–‡ä»¶åŸºæœ¬æ“ä½œå®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡æ–‡ä»¶çš„åŸºæœ¬è¯»å†™æ“ä½œ")
print(f"   âœ“ ç†è§£æ–‡ä»¶ç¼–ç å’Œå¼‚å¸¸å¤„ç†")
print(f"   âœ“ èƒ½å¤„ç†æ–‡ä»¶å±æ€§å’Œå…ƒæ•°æ®")
````

````python
print("ğŸ”§ æ–‡ä»¶ç³»ç»Ÿé«˜çº§æ“ä½œ:")
print("=" * 50)

import os
import shutil
import stat
import time
import zipfile
import json
from pathlib import Path
from datetime import datetime, timedelta

# 1. æ–‡ä»¶æƒé™ç®¡ç†
print(f"ğŸ“ 1. æ–‡ä»¶æƒé™ç®¡ç†:")

# 1.1 æŸ¥çœ‹æ–‡ä»¶æƒé™
print(f"\n   ğŸ” 1.1 æŸ¥çœ‹æ–‡ä»¶æƒé™:")

test_file = Path("test_files/permission_test.txt")
test_file.write_text("æƒé™æµ‹è¯•æ–‡ä»¶", encoding='utf-8')

# è·å–æ–‡ä»¶æƒé™
file_stat = test_file.stat()
file_mode = file_stat.st_mode

print(f"   æ–‡ä»¶: {test_file}")
print(f"   æƒé™å…«è¿›åˆ¶: {oct(file_mode)}")
print(f"   æƒé™åè¿›åˆ¶: {file_mode}")
print(f"   æ‰€æœ‰è€…è¯»: {bool(file_mode & stat.S_IRUSR)}")
print(f"   æ‰€æœ‰è€…å†™: {bool(file_mode & stat.S_IWUSR)}")
print(f"   æ‰€æœ‰è€…æ‰§è¡Œ: {bool(file_mode & stat.S_IXUSR)}")
print(f"   ç»„è¯»: {bool(file_mode & stat.S_IRGRP)}")
print(f"   ç»„å†™: {bool(file_mode & stat.S_IWGRP)}")
print(f"   ç»„æ‰§è¡Œ: {bool(file_mode & stat.S_IXGRP)}")
print(f"   å…¶ä»–è¯»: {bool(file_mode & stat.S_IROTH)}")
print(f"   å…¶ä»–å†™: {bool(file_mode & stat.S_IWOTH)}")
print(f"   å…¶ä»–æ‰§è¡Œ: {bool(file_mode & stat.S_IXOTH)}")

# 1.2 ä¿®æ”¹æ–‡ä»¶æƒé™ (Unixç³»ç»Ÿ)
print(f"\n   ğŸ”§ 1.2 ä¿®æ”¹æ–‡ä»¶æƒé™:")

def change_file_permissions(file_path, permissions):
    """ä¿®æ”¹æ–‡ä»¶æƒé™"""
    try:
        os.chmod(file_path, permissions)
        print(f"   æƒé™ä¿®æ”¹æˆåŠŸ: {oct(permissions)}")
        return True
    except Exception as e:
        print(f"   æƒé™ä¿®æ”¹å¤±è´¥: {e}")
        return False

# å°è¯•ä¿®æ”¹æƒé™ (åœ¨Windowsä¸Šå¯èƒ½å—é™)
if os.name == 'posix':  # Unixç³»ç»Ÿ
    # è®¾ç½®ä¸ºåªè¯»
    change_file_permissions(test_file, 0o444)
    
    # æ¢å¤è¯»å†™æƒé™
    change_file_permissions(test_file, 0o644)
else:
    print(f"   Windowsç³»ç»Ÿï¼Œæƒé™ä¿®æ”¹åŠŸèƒ½å—é™")

# 2. æ–‡ä»¶ç›‘æ§
print(f"\nğŸ“ 2. æ–‡ä»¶ç›‘æ§:")

# 2.1 ç®€å•çš„æ–‡ä»¶ç›‘æ§å™¨
print(f"\n   ğŸ‘€ 2.1 ç®€å•çš„æ–‡ä»¶ç›‘æ§å™¨:")

class SimpleFileWatcher:
    """ç®€å•çš„æ–‡ä»¶ç›‘æ§å™¨"""
    
    def __init__(self, directory):
        self.directory = Path(directory)
        self.file_states = {}
        self.callbacks = {
            'created': [],
            'modified': [],
            'deleted': []
        }
    
    def add_callback(self, event_type, callback):
        """æ·»åŠ äº‹ä»¶å›è°ƒ"""
        if event_type in self.callbacks:
            self.callbacks[event_type].append(callback)
    
    def _get_file_state(self, file_path):
        """è·å–æ–‡ä»¶çŠ¶æ€"""
        try:
            stat = file_path.stat()
            return {
                'size': stat.st_size,
                'mtime': stat.st_mtime,
                'exists': True
            }
        except FileNotFoundError:
            return {'exists': False}
    
    def _trigger_callbacks(self, event_type, file_path):
        """è§¦å‘å›è°ƒå‡½æ•°"""
        for callback in self.callbacks[event_type]:
            try:
                callback(file_path)
            except Exception as e:
                print(f"   å›è°ƒæ‰§è¡Œå¤±è´¥: {e}")
    
    def scan(self):
        """æ‰«æç›®å½•å˜åŒ–"""
        current_files = {}
        
        # æ‰«æå½“å‰ç›®å½•
        for file_path in self.directory.rglob("*"):
            if file_path.is_file():
                current_files[str(file_path)] = self._get_file_state(file_path)
        
        # æ£€æŸ¥æ–°æ–‡ä»¶å’Œä¿®æ”¹çš„æ–‡ä»¶
        for file_path, state in current_files.items():
            if file_path not in self.file_states:
                # æ–°æ–‡ä»¶
                self._trigger_callbacks('created', Path(file_path))
            elif self.file_states[file_path] != state:
                # æ–‡ä»¶ä¿®æ”¹
                self._trigger_callbacks('modified', Path(file_path))
        
        # æ£€æŸ¥åˆ é™¤çš„æ–‡ä»¶
        for file_path in self.file_states:
            if file_path not in current_files:
                self._trigger_callbacks('deleted', Path(file_path))
        
        # æ›´æ–°çŠ¶æ€
        self.file_states = current_files
    
    def watch(self, interval=1, duration=10):
        """ç›‘æ§ç›®å½•å˜åŒ–"""
        print(f"   å¼€å§‹ç›‘æ§ç›®å½•: {self.directory}")
        
        # åˆå§‹æ‰«æ
        self.scan()
        
        # ç›‘æ§å¾ªç¯
        start_time = time.time()
        while time.time() - start_time < duration:
            time.sleep(interval)
            self.scan()
        
        print(f"   ç›‘æ§ç»“æŸ")

# æµ‹è¯•æ–‡ä»¶ç›‘æ§å™¨
print(f"\n   æµ‹è¯•æ–‡ä»¶ç›‘æ§å™¨:")

# åˆ›å»ºç›‘æ§ç›®å½•
watch_dir = Path("test_files/watch_demo")
watch_dir.mkdir(exist_ok=True)

# åˆ›å»ºç›‘æ§å™¨
watcher = SimpleFileWatcher(watch_dir)

# æ·»åŠ äº‹ä»¶å¤„ç†å‡½æ•°
def on_file_created(file_path):
    print(f"     ğŸ“„ æ–‡ä»¶åˆ›å»º: {file_path.name}")

def on_file_modified(file_path):
    print(f"     âœï¸ æ–‡ä»¶ä¿®æ”¹: {file_path.name}")

def on_file_deleted(file_path):
    print(f"     ğŸ—‘ï¸ æ–‡ä»¶åˆ é™¤: {file_path.name}")

watcher.add_callback('created', on_file_created)
watcher.add_callback('modified', on_file_modified)
watcher.add_callback('deleted', on_file_deleted)

# æ¨¡æ‹Ÿæ–‡ä»¶æ“ä½œ
print(f"\n     æ¨¡æ‹Ÿæ–‡ä»¶æ“ä½œ:")
test_file1 = watch_dir / "test1.txt"
test_file2 = watch_dir / "test2.txt"

# åˆ›å»ºæ–‡ä»¶
test_file1.write_text("æµ‹è¯•æ–‡ä»¶1", encoding='utf-8')
time.sleep(0.1)

# ä¿®æ”¹æ–‡ä»¶
test_file1.write_text("æµ‹è¯•æ–‡ä»¶1 - ä¿®æ”¹ç‰ˆæœ¬", encoding='utf-8')
time.sleep(0.1)

# åˆ›å»ºå¦ä¸€ä¸ªæ–‡ä»¶
test_file2.write_text("æµ‹è¯•æ–‡ä»¶2", encoding='utf-8')
time.sleep(0.1)

# åˆ é™¤æ–‡ä»¶
test_file1.unlink()
time.sleep(0.1)

# æ‰«æå˜åŒ–
watcher.scan()

# 3. æ‰¹é‡æ–‡ä»¶æ“ä½œ
print(f"\nğŸ“ 3. æ‰¹é‡æ–‡ä»¶æ“ä½œ:")

# 3.1 æ‰¹é‡é‡å‘½å
print(f"\n   ğŸ“ 3.1 æ‰¹é‡é‡å‘½å:")

def batch_rename_files(directory, pattern, replacement):
    """æ‰¹é‡é‡å‘½åæ–‡ä»¶"""
    directory = Path(directory)
    renamed_files = []
    
    for file_path in directory.iterdir():
        if file_path.is_file() and pattern in file_path.name:
            new_name = file_path.name.replace(pattern, replacement)
            new_path = file_path.parent / new_name
            
            try:
                file_path.rename(new_path)
                renamed_files.append((file_path, new_path))
                print(f"     é‡å‘½å: {file_path.name} -> {new_name}")
            except Exception as e:
                print(f"     é‡å‘½åå¤±è´¥: {file_path.name}, é”™è¯¯: {e}")
    
    return renamed_files

# åˆ›å»ºæµ‹è¯•æ–‡ä»¶
batch_dir = Path("test_files/batch_demo")
batch_dir.mkdir(exist_ok=True)

for i in range(5):
    test_file = batch_dir / f"old_file_{i}.txt"
    test_file.write_text(f"æ–‡ä»¶ {i} çš„å†…å®¹", encoding='utf-8')

print(f"\n     æ‰¹é‡é‡å‘½åæµ‹è¯•:")
renamed = batch_rename_files(batch_dir, "old_file", "new_file")
print(f"     æˆåŠŸé‡å‘½å {len(renamed)} ä¸ªæ–‡ä»¶")

# 3.2 æ‰¹é‡æ–‡ä»¶å†…å®¹å¤„ç†
print(f"\n   ğŸ“ 3.2 æ‰¹é‡æ–‡ä»¶å†…å®¹å¤„ç†:")

def batch_process_files(directory, process_func, pattern="*.txt"):
    """æ‰¹é‡å¤„ç†æ–‡ä»¶å†…å®¹"""
    directory = Path(directory)
    processed_files = []
    
    for file_path in directory.glob(pattern):
        try:
            # è¯»å–åŸå†…å®¹
            original_content = file_path.read_text(encoding='utf-8')
            
            # å¤„ç†å†…å®¹
            processed_content = process_func(original_content)
            
            # å†™å…¥æ–°å†…å®¹
            file_path.write_text(processed_content, encoding='utf-8')
            
            processed_files.append(file_path)
            print(f"     å¤„ç†å®Œæˆ: {file_path.name}")
            
        except Exception as e:
            print(f"     å¤„ç†å¤±è´¥: {file_path.name}, é”™è¯¯: {e}")
    
    return processed_files

# å®šä¹‰å¤„ç†å‡½æ•°
def add_timestamp(content):
    """æ·»åŠ æ—¶é—´æˆ³"""
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    return f"å¤„ç†æ—¶é—´: {timestamp}\n{content}"

def add_line_numbers(content):
    """æ·»åŠ è¡Œå·"""
    lines = content.split('\n')
    numbered_lines = [f"{i+1:2d}: {line}" for i, line in enumerate(lines)]
    return '\n'.join(numbered_lines)

# æ‰¹é‡å¤„ç†æ–‡ä»¶
print(f"\n     æ‰¹é‡æ·»åŠ æ—¶é—´æˆ³:")
processed = batch_process_files(batch_dir, add_timestamp)
print(f"     å¤„ç†äº† {len(processed)} ä¸ªæ–‡ä»¶")

print(f"\n     æ‰¹é‡æ·»åŠ è¡Œå·:")
processed = batch_process_files(batch_dir, add_line_numbers)
print(f"     å¤„ç†äº† {len(processed)} ä¸ªæ–‡ä»¶")

# 4. æ–‡ä»¶å‹ç¼©å’Œè§£å‹
print(f"\nğŸ“ 4. æ–‡ä»¶å‹ç¼©å’Œè§£å‹:")

# 4.1 åˆ›å»ºZIPæ–‡ä»¶
print(f"\n   ğŸ“¦ 4.1 åˆ›å»ºZIPæ–‡ä»¶:")

def create_zip_archive(source_dir, zip_path):
    """åˆ›å»ºZIPå‹ç¼©æ–‡ä»¶"""
    source_dir = Path(source_dir)
    zip_path = Path(zip_path)
    
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for file_path in source_dir.rglob("*"):
            if file_path.is_file():
                # è®¡ç®—ç›¸å¯¹è·¯å¾„
                arcname = file_path.relative_to(source_dir.parent)
                zipf.write(file_path, arcname)
                print(f"     æ·»åŠ åˆ°å‹ç¼©åŒ…: {arcname}")
    
    print(f"   å‹ç¼©å®Œæˆ: {zip_path}")
    return zip_path

# åˆ›å»ºå‹ç¼©æ–‡ä»¶
zip_file = Path("test_files/demo_archive.zip")
create_zip_archive(batch_dir, zip_file)

# 4.2 è§£å‹ZIPæ–‡ä»¶
print(f"\n   ğŸ“‚ 4.2 è§£å‹ZIPæ–‡ä»¶:")

def extract_zip_archive(zip_path, extract_dir):
    """è§£å‹ZIPæ–‡ä»¶"""
    zip_path = Path(zip_path)
    extract_dir = Path(extract_dir)
    extract_dir.mkdir(exist_ok=True)
    
    with zipfile.ZipFile(zip_path, 'r') as zipf:
        file_list = zipf.namelist()
        print(f"   å‹ç¼©åŒ…å†…å®¹: {len(file_list)} ä¸ªæ–‡ä»¶")
        
        for file_name in file_list:
            zipf.extract(file_name, extract_dir)
            print(f"     è§£å‹: {file_name}")
    
    print(f"   è§£å‹å®Œæˆ: {extract_dir}")
    return extract_dir

# è§£å‹æ–‡ä»¶
extract_dir = Path("test_files/extracted")
extract_zip_archive(zip_file, extract_dir)

# 5. æ–‡ä»¶æ¸…ç†å·¥å…·
print(f"\nğŸ“ 5. æ–‡ä»¶æ¸…ç†å·¥å…·:")

class FileCleaner:
    """æ–‡ä»¶æ¸…ç†å·¥å…·"""
    
    def __init__(self, directory):
        self.directory = Path(directory)
    
    def clean_empty_files(self, min_size=0):
        """æ¸…ç†ç©ºæ–‡ä»¶æˆ–å°äºæŒ‡å®šå¤§å°çš„æ–‡ä»¶"""
        removed_files = []
        
        for file_path in self.directory.rglob("*"):
            if file_path.is_file():
                file_size = file_path.stat().st_size
                if file_size <= min_size:
                    file_path.unlink()
                    removed_files.append(file_path)
                    print(f"     åˆ é™¤ç©ºæ–‡ä»¶: {file_path.name}")
        
        return removed_files
    
    def clean_old_files(self, days_old=30):
        """æ¸…ç†æ—§æ–‡ä»¶"""
        removed_files = []
        cutoff_time = time.time() - (days_old * 24 * 3600)
        
        for file_path in self.directory.rglob("*"):
            if file_path.is_file():
                file_mtime = file_path.stat().st_mtime
                if file_mtime < cutoff_time:
                    file_path.unlink()
                    removed_files.append(file_path)
                    print(f"     åˆ é™¤æ—§æ–‡ä»¶: {file_path.name}")
        
        return removed_files
    
    def clean_temp_files(self, patterns=None):
        """æ¸…ç†ä¸´æ—¶æ–‡ä»¶"""
        if patterns is None:
            patterns = ['*.tmp', '*.temp', '*~', '*.bak', '.DS_Store']
        
        removed_files = []
        
        for pattern in patterns:
            for file_path in self.directory.rglob(pattern):
                if file_path.is_file():
                    file_path.unlink()
                    removed_files.append(file_path)
                    print(f"     åˆ é™¤ä¸´æ—¶æ–‡ä»¶: {file_path.name}")
        
        return removed_files
    
    def organize_files_by_extension(self):
        """æŒ‰æ‰©å±•åç»„ç»‡æ–‡ä»¶"""
        organized_files = []
        
        for file_path in self.directory.iterdir():
            if file_path.is_file():
                extension = file_path.suffix.lower()
                if extension:
                    # åˆ›å»ºæ‰©å±•åç›®å½•
                    ext_dir = self.directory / extension[1:]  # å»æ‰ç‚¹å·
                    ext_dir.mkdir(exist_ok=True)
                    
                    # ç§»åŠ¨æ–‡ä»¶
                    new_path = ext_dir / file_path.name
                    shutil.move(str(file_path), str(new_path))
                    organized_files.append((file_path, new_path))
                    print(f"     ç»„ç»‡æ–‡ä»¶: {file_path.name} -> {extension[1:]}/")
        
        return organized_files

# æµ‹è¯•æ–‡ä»¶æ¸…ç†å·¥å…·
print(f"\n   æµ‹è¯•æ–‡ä»¶æ¸…ç†å·¥å…·:")

# åˆ›å»ºä¸€äº›æµ‹è¯•æ–‡ä»¶
clean_dir = Path("test_files/clean_demo")
clean_dir.mkdir(exist_ok=True)

# åˆ›å»ºä¸åŒç±»å‹çš„æ–‡ä»¶
(clean_dir / "empty.txt").touch()
(clean_dir / "small.txt").write_text("x", encoding='utf-8')
(clean_dir / "normal.txt").write_text("æ­£å¸¸æ–‡ä»¶å†…å®¹", encoding='utf-8')
(clean_dir / "document.pdf").write_text("PDFå†…å®¹", encoding='utf-8')
(clean_dir / "image.jpg").write_text("å›¾ç‰‡å†…å®¹", encoding='utf-8')
(clean_dir / "temp.tmp").write_text("ä¸´æ—¶æ–‡ä»¶", encoding='utf-8')
(clean_dir / "backup.bak").write_text("å¤‡ä»½æ–‡ä»¶", encoding='utf-8')

# åˆ›å»ºæ¸…ç†å™¨
cleaner = FileCleaner(clean_dir)

# æ¸…ç†ç©ºæ–‡ä»¶
print(f"\n     æ¸…ç†ç©ºæ–‡ä»¶:")
empty_removed = cleaner.clean_empty_files(min_size=1)
print(f"     åˆ é™¤äº† {len(empty_removed)} ä¸ªç©ºæ–‡ä»¶")

# æ¸…ç†ä¸´æ—¶æ–‡ä»¶
print(f"\n     æ¸…ç†ä¸´æ—¶æ–‡ä»¶:")
temp_removed = cleaner.clean_temp_files()
print(f"     åˆ é™¤äº† {len(temp_removed)} ä¸ªä¸´æ—¶æ–‡ä»¶")

# æŒ‰æ‰©å±•åç»„ç»‡æ–‡ä»¶
print(f"\n     æŒ‰æ‰©å±•åç»„ç»‡æ–‡ä»¶:")
organized = cleaner.organize_files_by_extension()
print(f"     ç»„ç»‡äº† {len(organized)} ä¸ªæ–‡ä»¶")

print(f"\nâœ… æ–‡ä»¶ç³»ç»Ÿé«˜çº§æ“ä½œå®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡æ–‡ä»¶æƒé™å’Œå±æ€§ç®¡ç†")
print(f"   âœ“ ç†è§£æ–‡ä»¶ç›‘æ§å’Œäº‹ä»¶å¤„ç†")
print(f"   âœ“ èƒ½è¿›è¡Œæ‰¹é‡æ–‡ä»¶æ“ä½œ")
print(f"   âœ“ æŒæ¡å‹ç¼©å’Œè§£å‹ç¼©æŠ€æœ¯")
````
