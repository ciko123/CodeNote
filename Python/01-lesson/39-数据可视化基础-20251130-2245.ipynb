````python
print("ğŸ¨ Seabornç»Ÿè®¡å›¾è¡¨:")
print("=" * 50)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from typing import List, Tuple, Any, Optional, Union, Dict
from dataclasses import dataclass
import json
from datetime import datetime, timedelta

# è®¾ç½®Seabornæ ·å¼
sns.set_palette("husl")  # è®¾ç½®è°ƒè‰²æ¿
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 1. SeabornåŸºç¡€å›¾è¡¨
print(f"ğŸ“ 1. SeabornåŸºç¡€å›¾è¡¨:")

# 1.1 åˆ†å¸ƒå›¾
print(f"\n   ğŸ“Š 1.1 åˆ†å¸ƒå›¾:")

def demonstrate_distribution_plots():
    """æ¼”ç¤ºåˆ†å¸ƒå›¾"""
    
    # åˆ›å»ºç¤ºä¾‹æ•°æ®
    np.random.seed(42)
    data = {
        'normal': np.random.normal(100, 15, 1000),
        'exponential': np.random.exponential(2, 1000),
        'uniform': np.random.uniform(0, 10, 1000)
    }
    
    df_dist = pd.DataFrame(data)
    
    # 1. ç›´æ–¹å›¾ + KDE
    print(f"   åˆ›å»ºç›´æ–¹å›¾ + KDE...")
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    sns.histplot(data=df_dist['normal'], kde=True, color='skyblue')
    plt.title('æ­£æ€åˆ†å¸ƒ', fontweight='bold')
    
    plt.subplot(1, 3, 2)
    sns.histplot(data=df_dist['exponential'], kde=True, color='lightcoral')
    plt.title('æŒ‡æ•°åˆ†å¸ƒ', fontweight='bold')
    
    plt.subplot(1, 3, 3)
    sns.histplot(data=df_dist['uniform'], kde=True, color='lightgreen')
    plt.title('å‡åŒ€åˆ†å¸ƒ', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    print(f"   âœ… ç›´æ–¹å›¾ + KDE åˆ›å»ºå®Œæˆ")
    
    # 2. æ ¸å¯†åº¦ä¼°è®¡å›¾
    print(f"\n   åˆ›å»ºæ ¸å¯†åº¦ä¼°è®¡å›¾...")
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=df_dist, fill=True, alpha=0.7)
    plt.title('å¤šå˜é‡æ ¸å¯†åº¦ä¼°è®¡å›¾', fontsize=16, fontweight='bold')
    plt.xlabel('æ•°å€¼', fontsize=12)
    plt.ylabel('å¯†åº¦', fontsize=12)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… æ ¸å¯†åº¦ä¼°è®¡å›¾åˆ›å»ºå®Œæˆ")
    
    # 3. ç®±çº¿å›¾
    print(f"\n   åˆ›å»ºç®±çº¿å›¾...")
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df_dist)
    plt.title('åˆ†å¸ƒç®±çº¿å›¾', fontsize=16, fontweight='bold')
    plt.ylabel('æ•°å€¼', fontsize=12)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… ç®±çº¿å›¾åˆ›å»ºå®Œæˆ")
    
    # 4. å°æç´å›¾
    print(f"\n   åˆ›å»ºå°æç´å›¾...")
    plt.figure(figsize=(10, 6))
    sns.violinplot(data=df_dist)
    plt.title('åˆ†å¸ƒå°æç´å›¾', fontsize=16, fontweight='bold')
    plt.ylabel('æ•°å€¼', fontsize=12)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… å°æç´å›¾åˆ›å»ºå®Œæˆ")

demonstrate_distribution_plots()

# 1.2 å…³ç³»å›¾
print(f"\n   ğŸ”— 1.2 å…³ç³»å›¾:")

def demonstrate_relationship_plots():
    """æ¼”ç¤ºå…³ç³»å›¾"""
    
    # åˆ›å»ºç¤ºä¾‹æ•°æ®
    np.random.seed(42)
    n_samples = 200
    
    data_rel = {
        'x': np.random.randn(n_samples),
        'y': np.random.randn(n_samples),
        'size': np.random.rand(n_samples) * 100,
        'category': np.random.choice(['A', 'B', 'C'], n_samples),
        'value': np.random.randn(n_samples) * 10 + 50
    }
    
    df_rel = pd.DataFrame(data_rel)
    
    # 1. æ•£ç‚¹å›¾
    print(f"   åˆ›å»ºæ•£ç‚¹å›¾...")
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=df_rel, x='x', y='y', hue='category', size='size', 
                    sizes=(20, 200), alpha=0.7)
    plt.title('å¤šç»´æ•£ç‚¹å›¾', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    print(f"   âœ… æ•£ç‚¹å›¾åˆ›å»ºå®Œæˆ")
    
    # 2. çº¿æ€§å›å½’å›¾
    print(f"\n   åˆ›å»ºçº¿æ€§å›å½’å›¾...")
    plt.figure(figsize=(10, 6))
    sns.regplot(data=df_rel, x='x', y='y', scatter_kws={'alpha':0.6}, 
                line_kws={'color': 'red', 'linewidth': 2})
    plt.title('çº¿æ€§å›å½’å›¾', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    print(f"   âœ… çº¿æ€§å›å½’å›¾åˆ›å»ºå®Œæˆ")
    
    # 3. è”åˆåˆ†å¸ƒå›¾
    print(f"\n   åˆ›å»ºè”åˆåˆ†å¸ƒå›¾...")
    sns.jointplot(data=df_rel, x='x', y='y', kind='scatter', 
                  hue='category', height=8)
    plt.suptitle('è”åˆåˆ†å¸ƒå›¾', fontsize=16, fontweight='bold', y=1.02)
    plt.show()
    print(f"   âœ… è”åˆåˆ†å¸ƒå›¾åˆ›å»ºå®Œæˆ")
    
    # 4. é…å¯¹å›¾
    print(f"\n   åˆ›å»ºé…å¯¹å›¾...")
    # é€‰æ‹©æ•°å€¼åˆ—åˆ›å»ºé…å¯¹å›¾
    numeric_cols = ['x', 'y', 'value']
    sns.pairplot(df_rel[numeric_cols], height=2.5)
    plt.suptitle('æ•°å€¼å˜é‡é…å¯¹å›¾', fontsize=16, fontweight='bold', y=1.02)
    plt.show()
    print(f"   âœ… é…å¯¹å›¾åˆ›å»ºå®Œæˆ")

demonstrate_relationship_plots()

# 1.3 åˆ†ç±»å›¾
print(f"\n   ğŸ“‚ 1.3 åˆ†ç±»å›¾:")

def demonstrate_categorical_plots():
    """æ¼”ç¤ºåˆ†ç±»å›¾"""
    
    # åˆ›å»ºç¤ºä¾‹æ•°æ®
    np.random.seed(42)
    categories = ['æŠ€æœ¯', 'åˆ›æ„', 'åˆ†æ', 'é€šç”¨']
    models = ['GPT-3.5', 'GPT-4', 'Claude', 'LLaMA']
    
    data_cat = []
    for _ in range(300):
        data_cat.append({
            'category': np.random.choice(categories),
            'model': np.random.choice(models),
            'response_time': np.random.randint(500, 3000),
            'satisfaction': np.random.uniform(3.0, 5.0),
            'tokens': np.random.randint(20, 150)
        })
    
    df_cat = pd.DataFrame(data_cat)
    
    # 1. åˆ†ç±»æ•£ç‚¹å›¾
    print(f"   åˆ›å»ºåˆ†ç±»æ•£ç‚¹å›¾...")
    plt.figure(figsize=(12, 6))
    sns.stripplot(data=df_cat, x='category', y='response_time', hue='model', 
                   jitter=True, dodge=True, alpha=0.7)
    plt.title('åˆ†ç±»æ•£ç‚¹å›¾', fontsize=16, fontweight='bold')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… åˆ†ç±»æ•£ç‚¹å›¾åˆ›å»ºå®Œæˆ")
    
    # 2. ç®±çº¿å›¾
    print(f"\n   åˆ›å»ºåˆ†ç±»ç®±çº¿å›¾...")
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df_cat, x='category', y='satisfaction', hue='model')
    plt.title('åˆ†ç±»ç®±çº¿å›¾', fontsize=16, fontweight='bold')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… åˆ†ç±»ç®±çº¿å›¾åˆ›å»ºå®Œæˆ")
    
    # 3. å°æç´å›¾
    print(f"\n   åˆ›å»ºåˆ†ç±»å°æç´å›¾...")
    plt.figure(figsize=(12, 6))
    sns.violinplot(data=df_cat, x='category', y='tokens', hue='model', 
                   split=True, inner='quartile')
    plt.title('åˆ†ç±»å°æç´å›¾', fontsize=16, fontweight='bold')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… åˆ†ç±»å°æç´å›¾åˆ›å»ºå®Œæˆ")
    
    # 4. çƒ­åŠ›å›¾
    print(f"\n   åˆ›å»ºçƒ­åŠ›å›¾...")
    # åˆ›å»ºé€è§†è¡¨æ•°æ®
    pivot_table = df_cat.pivot_table(values='satisfaction', 
                                     index='category', 
                                     columns='model', 
                                     aggfunc='mean')
    
    plt.figure(figsize=(10, 6))
    sns.heatmap(pivot_table, annot=True, cmap='YlOrRd', fmt='.2f', 
                cbar_kws={'label': 'å¹³å‡æ»¡æ„åº¦'})
    plt.title('ç±»åˆ«-æ¨¡å‹æ»¡æ„åº¦çƒ­åŠ›å›¾', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    print(f"   âœ… çƒ­åŠ›å›¾åˆ›å»ºå®Œæˆ")

demonstrate_categorical_plots()

# 2. LangChainç»Ÿè®¡å¯è§†åŒ–åº”ç”¨
print(f"\nğŸ¤– 2. LangChainç»Ÿè®¡å¯è§†åŒ–åº”ç”¨:")

@dataclass
class StatisticalAnalyzer:
    """ç»Ÿè®¡åˆ†æå™¨ - æ¨¡æ‹ŸLangChainçš„ç»Ÿè®¡åˆ†æå¯è§†åŒ–"""
    
    def __init__(self):
        self.data = None
    
    def generate_analysis_data(self, num_records: int = 500):
        """ç”Ÿæˆåˆ†ææ•°æ®"""
        np.random.seed(42)
        
        users = [f'user_{i:03d}' for i in range(1, 21)]
        models = ['gpt-3.5-turbo', 'gpt-4', 'claude-2', 'llama-2']
        categories = ['æŠ€æœ¯', 'åˆ›æ„', 'åˆ†æ', 'é€šç”¨', 'æ•™è‚²']
        
        data = []
        for i in range(num_records):
            record = {
                'user_id': np.random.choice(users),
                'model': np.random.choice(models),
                'category': np.random.choice(categories),
                'tokens_used': np.random.randint(10, 200),
                'response_time_ms': np.random.randint(300, 4000),
                'satisfaction_score': np.random.uniform(2.5, 5.0),
                'cost_usd': np.random.uniform(0.01, 0.25),
                'success': np.random.choice([True, False], p=[0.85, 0.15]),
                'hour': np.random.randint(0, 24),
                'day_of_week': np.random.choice(['å‘¨ä¸€', 'å‘¨äºŒ', 'å‘¨ä¸‰', 'å‘¨å››', 'å‘¨äº”', 'å‘¨å…­', 'å‘¨æ—¥'])
            }
            data.append(record)
        
        self.data = pd.DataFrame(data)
    
    def create_comprehensive_dashboard(self):
        """åˆ›å»ºç»¼åˆä»ªè¡¨æ¿"""
        if self.data is None:
            print("   è¯·å…ˆç”Ÿæˆæ•°æ®")
            return
        
        fig = plt.figure(figsize=(20, 16))
        
        # 1. ä½¿ç”¨æ—¶é—´åˆ†å¸ƒ
        ax1 = plt.subplot(3, 4, 1)
        hourly_data = self.data.groupby('hour').size()
        sns.lineplot(x=hourly_data.index, y=hourly_data.values, ax=ax1)
        ax1.set_title('24å°æ—¶ä½¿ç”¨åˆ†å¸ƒ', fontweight='bold')
        ax1.set_xlabel('å°æ—¶')
        ax1.set_ylabel('å¯¹è¯æ•°')
        
        # 2. æ»¡æ„åº¦åˆ†å¸ƒ
        ax2 = plt.subplot(3, 4, 2)
        sns.histplot(data=self.data, x='satisfaction_score', kde=True, ax=ax2)
        ax2.set_title('æ»¡æ„åº¦åˆ†å¸ƒ', fontweight='bold')
        ax2.set_xlabel('æ»¡æ„åº¦')
        
        # 3. æ¨¡å‹ä½¿ç”¨å¯¹æ¯”
        ax3 = plt.subplot(3, 4, 3)
        model_counts = self.data['model'].value_counts()
        sns.barplot(x=model_counts.index, y=model_counts.values, ax=ax3)
        ax3.set_title('æ¨¡å‹ä½¿ç”¨é¢‘æ¬¡', fontweight='bold')
        ax3.set_xlabel('æ¨¡å‹')
        ax3.set_ylabel('ä½¿ç”¨æ¬¡æ•°')
        ax3.tick_params(axis='x', rotation=45)
        
        # 4. å“åº”æ—¶é—´åˆ†å¸ƒ
        ax4 = plt.subplot(3, 4, 4)
        sns.boxplot(data=self.data, x='model', y='response_time_ms', ax=ax4)
        ax4.set_title('å“åº”æ—¶é—´åˆ†å¸ƒ', fontweight='bold')
        ax4.set_xlabel('æ¨¡å‹')
        ax4.set_ylabel('å“åº”æ—¶é—´(ms)')
        ax4.tick_params(axis='x', rotation=45)
        
        # 5. æˆåŠŸç‡é¥¼å›¾
        ax5 = plt.subplot(3, 4, 5)
        success_counts = self.data['success'].value_counts()
        ax5.pie(success_counts.values, labels=['æˆåŠŸ', 'å¤±è´¥'], autopct='%1.1f%%', 
                colors=['#4CAF50', '#F44336'])
        ax5.set_title('æˆåŠŸç‡', fontweight='bold')
        
        # 6. ç±»åˆ«ä½¿ç”¨çƒ­åŠ›å›¾
        ax6 = plt.subplot(3, 4, 6)
        category_model = self.data.pivot_table(values='tokens_used', 
                                              index='category', 
                                              columns='model', 
                                              aggfunc='mean')
        sns.heatmap(category_model, annot=True, fmt='.0f', cmap='Blues', ax=ax6)
        ax6.set_title('ç±»åˆ«-æ¨¡å‹å¹³å‡Token', fontweight='bold')
        
        # 7. æˆæœ¬ vs æ»¡æ„åº¦æ•£ç‚¹å›¾
        ax7 = plt.subplot(3, 4, 7)
        sns.scatterplot(data=self.data, x='cost_usd', y='satisfaction_score', 
                       hue='model', alpha=0.6, ax=ax7)
        ax7.set_title('æˆæœ¬ vs æ»¡æ„åº¦', fontweight='bold')
        ax7.set_xlabel('æˆæœ¬(USD)')
        ax7.set_ylabel('æ»¡æ„åº¦')
        
        # 8. æ˜ŸæœŸåˆ†å¸ƒ
        ax8 = plt.subplot(3, 4, 8)
        day_counts = self.data['day_of_week'].value_counts()
        sns.barplot(x=day_counts.index, y=day_counts.values, ax=ax8)
        ax8.set_title('æ˜ŸæœŸåˆ†å¸ƒ', fontweight='bold')
        ax8.set_xlabel('æ˜ŸæœŸ')
        ax8.set_ylabel('å¯¹è¯æ•°')
        ax8.tick_params(axis='x', rotation=45)
        
        # 9. Tokenä½¿ç”¨åˆ†å¸ƒ
        ax9 = plt.subplot(3, 4, 9)
        sns.violinplot(data=self.data, x='category', y='tokens_used', ax=ax9)
        ax9.set_title('Tokenä½¿ç”¨åˆ†å¸ƒ', fontweight='bold')
        ax9.set_xlabel('ç±»åˆ«')
        ax9.set_ylabel('Tokenæ•°')
        ax9.tick_params(axis='x', rotation=45)
        
        # 10. æ¨¡å‹æ€§èƒ½å¯¹æ¯”
        ax10 = plt.subplot(3, 4, 10)
        model_perf = self.data.groupby('model').agg({
            'satisfaction_score': 'mean',
            'response_time_ms': 'mean',
            'cost_usd': 'mean'
        }).round(2)
        
        # æ ‡å‡†åŒ–æ•°æ®ç”¨äºå¯¹æ¯”
        perf_normalized = (model_perf - model_perf.min()) / (model_perf.max() - model_perf.min())
        sns.heatmap(perf_normalized.T, annot=True, cmap='RdYlGn', ax=ax10)
        ax10.set_title('æ¨¡å‹æ€§èƒ½å¯¹æ¯”(æ ‡å‡†åŒ–)', fontweight='bold')
        
        # 11. ç”¨æˆ·æ´»è·ƒåº¦
        ax11 = plt.subplot(3, 4, 11)
        user_activity = self.data['user_id'].value_counts().head(10)
        sns.barplot(y=user_activity.index, x=user_activity.values, ax=ax11, orient='h')
        ax11.set_title('Top 10 æ´»è·ƒç”¨æˆ·', fontweight='bold')
        ax11.set_xlabel('å¯¹è¯æ•°')
        ax11.set_ylabel('ç”¨æˆ·ID')
        
        # 12. ç»¼åˆè¯„åˆ†é›·è¾¾å›¾æ•°æ®å‡†å¤‡
        ax12 = plt.subplot(3, 4, 12, projection='polar')
        model_stats = self.data.groupby('model').agg({
            'satisfaction_score': 'mean',
            'response_time_ms': lambda x: 1 - (x - x.min()) / (x.max() - x.min()),  # åå‘æ ‡å‡†åŒ–å“åº”æ—¶é—´
            'cost_usd': lambda x: 1 - (x - x.min()) / (x.max() - x.min()),  # åå‘æ ‡å‡†åŒ–æˆæœ¬
            'success': 'mean'
        }).round(2)
        
        # ç®€åŒ–çš„é›·è¾¾å›¾å±•ç¤º
        categories_radar = ['æ»¡æ„åº¦', 'é€Ÿåº¦', 'ç»æµæ€§', 'æˆåŠŸç‡']
        angles = np.linspace(0, 2 * np.pi, len(categories_radar), endpoint=False).tolist()
        angles += angles[:1]  # é—­åˆå›¾å½¢
        
        for i, model in enumerate(model_stats.index):
            values = model_stats.loc[model].values
            values = values.tolist() + values[:1]  # é—­åˆå›¾å½¢
            ax12.plot(angles, values, 'o-', linewidth=2, label=model)
            ax12.fill(angles, values, alpha=0.25)
        
        ax12.set_xticks(angles[:-1])
        ax12.set_xticklabels(categories_radar)
        ax12.set_title('æ¨¡å‹ç»¼åˆè¯„åˆ†', fontweight='bold', pad=20)
        ax12.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
        
        plt.tight_layout()
        plt.show()
        
        return {
            'total_records': len(self.data),
            'unique_users': self.data['user_id'].nunique(),
            'avg_satisfaction': self.data['satisfaction_score'].mean(),
            'success_rate': self.data['success'].mean(),
            'total_cost': self.data['cost_usd'].sum()
        }

def demonstrate_statistical_analyzer():
    """æ¼”ç¤ºç»Ÿè®¡åˆ†æå™¨"""
    analyzer = StatisticalAnalyzer()
    
    # ç”Ÿæˆåˆ†ææ•°æ®
    print(f"   ç»Ÿè®¡åˆ†æå¯è§†åŒ–æ¼”ç¤º:")
    analyzer.generate_analysis_data(500)
    print(f"   ç”Ÿæˆäº† {len(analyzer.data)} æ¡åˆ†ææ•°æ®")
    
    # åˆ›å»ºç»¼åˆä»ªè¡¨æ¿
    print(f"\n   åˆ›å»ºç»¼åˆä»ªè¡¨æ¿...")
    summary = analyzer.create_comprehensive_dashboard()
    print(f"   âœ… ç»¼åˆä»ªè¡¨æ¿åˆ›å»ºå®Œæˆ")
    
    # æ˜¾ç¤ºå…³é”®æŒ‡æ ‡
    print(f"\n   ğŸ“Š å…³é”®æŒ‡æ ‡æ‘˜è¦:")
    for key, value in summary.items():
        if isinstance(value, float):
            print(f"   {key}: {value:.2f}")
        else:
            print(f"   {key}: {value}")

demonstrate_statistical_analyzer()

print(f"\nâœ… Seabornç»Ÿè®¡å›¾è¡¨å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡SeabornåŸºç¡€ç»˜å›¾æ–¹æ³•")
print(f"   âœ“ ç†è§£ç»Ÿè®¡å›¾è¡¨çš„ç±»å‹å’Œç”¨é€”")
print(f"   âœ“ ç†Ÿç»ƒä½¿ç”¨Seabornçš„æ ·å¼å’Œä¸»é¢˜")
print(f"   âœ“ èƒ½å¤Ÿåˆ›å»ºå¤æ‚çš„ç»Ÿè®¡å¯è§†åŒ–")
````

````python
print("ğŸ¼ Pandaså†…ç½®å¯è§†åŒ–:")
print("=" * 50)

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from typing import List, Tuple, Any, Optional, Union, Dict
from dataclasses import dataclass
import json
from datetime import datetime, timedelta

# è®¾ç½®æ ·å¼
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('seaborn-v0_8')

# 1. PandasåŸºç¡€å¯è§†åŒ–
print(f"ğŸ“ 1. PandasåŸºç¡€å¯è§†åŒ–:")

# 1.1 Serieså¯è§†åŒ–
print(f"\n   ğŸ“Š 1.1 Serieså¯è§†åŒ–:")

def demonstrate_series_plotting():
    """æ¼”ç¤ºSerieså¯è§†åŒ–"""
    
    # åˆ›å»ºæ—¶é—´åºåˆ—æ•°æ®
    dates = pd.date_range('2023-01-01', periods=100, freq='D')
    values = np.random.randn(100).cumsum() + 100
    ts = pd.Series(values, index=dates, name='è‚¡ç¥¨ä»·æ ¼')
    
    # 1. çº¿å›¾
    print(f"   åˆ›å»ºSeriesçº¿å›¾...")
    plt.figure(figsize=(12, 6))
    ts.plot(linewidth=2, color='blue', alpha=0.8)
    plt.title('æ—¶é—´åºåˆ—çº¿å›¾', fontsize=16, fontweight='bold')
    plt.xlabel('æ—¥æœŸ', fontsize=12)
    plt.ylabel('ä»·æ ¼', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… Seriesçº¿å›¾åˆ›å»ºå®Œæˆ")
    
    # 2. æŸ±çŠ¶å›¾
    print(f"\n   åˆ›å»ºSeriesæŸ±çŠ¶å›¾...")
    categories = pd.Series(
        [45, 78, 56, 89, 34, 67, 90, 23, 56, 78],
        index=['äº§å“A', 'äº§å“B', 'äº§å“C', 'äº§å“D', 'äº§å“E', 
               'äº§å“F', 'äº§å“G', 'äº§å“H', 'äº§å“I', 'äº§å“J'],
        name='é”€å”®é¢'
    )
    
    plt.figure(figsize=(12, 6))
    categories.plot(kind='bar', color='skyblue', alpha=0.8)
    plt.title('äº§å“é”€å”®é¢æŸ±çŠ¶å›¾', fontsize=16, fontweight='bold')
    plt.xlabel('äº§å“', fontsize=12)
    plt.ylabel('é”€å”®é¢', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… SeriesæŸ±çŠ¶å›¾åˆ›å»ºå®Œæˆ")
    
    # 3. é¥¼å›¾
    print(f"\n   åˆ›å»ºSeriesé¥¼å›¾...")
    market_share = pd.Series(
        [35, 25, 20, 15, 5],
        index=['å“ç‰ŒA', 'å“ç‰ŒB', 'å“ç‰ŒC', 'å“ç‰ŒD', 'å…¶ä»–'],
        name='å¸‚åœºä»½é¢'
    )
    
    plt.figure(figsize=(10, 8))
    market_share.plot(kind='pie', autopct='%1.1f%%', shadow=True, 
                      startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc'])
    plt.title('å¸‚åœºä»½é¢é¥¼å›¾', fontsize=16, fontweight='bold')
    plt.ylabel('')  # éšè—yè½´æ ‡ç­¾
    plt.tight_layout()
    plt.show()
    print(f"   âœ… Seriesé¥¼å›¾åˆ›å»ºå®Œæˆ")
    
    # 4. ç›´æ–¹å›¾
    print(f"\n   åˆ›å»ºSeriesç›´æ–¹å›¾...")
    scores = pd.Series(np.random.normal(75, 10, 1000), name='è€ƒè¯•æˆç»©')
    
    plt.figure(figsize=(10, 6))
    scores.plot(kind='hist', bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
    plt.title('æˆç»©åˆ†å¸ƒç›´æ–¹å›¾', fontsize=16, fontweight='bold')
    plt.xlabel('åˆ†æ•°', fontsize=12)
    plt.ylabel('é¢‘æ¬¡', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… Seriesç›´æ–¹å›¾åˆ›å»ºå®Œæˆ")

demonstrate_series_plotting()

# 1.2 DataFrameå¯è§†åŒ–
print(f"\n   ğŸ—‚ï¸ 1.2 DataFrameå¯è§†åŒ–:")

def demonstrate_dataframe_plotting():
    """æ¼”ç¤ºDataFrameå¯è§†åŒ–"""
    
    # åˆ›å»ºç¤ºä¾‹DataFrame
    dates = pd.date_range('2023-01-01', periods=50, freq='D')
    data = {
        'äº§å“A': np.random.randn(50).cumsum() + 100,
        'äº§å“B': np.random.randn(50).cumsum() + 80,
        'äº§å“C': np.random.randn(50).cumsum() + 120,
        'äº§å“D': np.random.randn(50).cumsum() + 90
    }
    df = pd.DataFrame(data, index=dates)
    
    # 1. å¤šçº¿å›¾
    print(f"   åˆ›å»ºDataFrameå¤šçº¿å›¾...")
    plt.figure(figsize=(12, 6))
    df.plot(linewidth=2, alpha=0.8)
    plt.title('å¤šäº§å“ä»·æ ¼è¶‹åŠ¿', fontsize=16, fontweight='bold')
    plt.xlabel('æ—¥æœŸ', fontsize=12)
    plt.ylabel('ä»·æ ¼', fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… DataFrameå¤šçº¿å›¾åˆ›å»ºå®Œæˆ")
    
    # 2. å­å›¾
    print(f"\n   åˆ›å»ºDataFrameå­å›¾...")
    df.plot(subplots=True, figsize=(15, 10), linewidth=2, alpha=0.8)
    plt.suptitle('å„äº§å“ä»·æ ¼è¶‹åŠ¿å­å›¾', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    print(f"   âœ… DataFrameå­å›¾åˆ›å»ºå®Œæˆ")
    
    # 3. æŸ±çŠ¶å›¾
    print(f"\n   åˆ›å»ºDataFrameæŸ±çŠ¶å›¾...")
    sales_data = {
        'Q1': [120, 150, 180, 200],
        'Q2': [130, 160, 170, 210],
        'Q3': [140, 155, 190, 220],
        'Q4': [150, 165, 200, 230]
    }
    df_sales = pd.DataFrame(sales_data, index=['äº§å“A', 'äº§å“B', 'äº§å“C', 'äº§å“D'])
    
    plt.figure(figsize=(12, 6))
    df_sales.plot(kind='bar', width=0.8)
    plt.title('å­£åº¦é”€å”®å¯¹æ¯”', fontsize=16, fontweight='bold')
    plt.xlabel('äº§å“', fontsize=12)
    plt.ylabel('é”€å”®é¢', fontsize=12)
    plt.xticks(rotation=0)
    plt.legend(title='å­£åº¦')
    plt.tight_layout()
    plt.show()
    print(f"   âœ… DataFrameæŸ±çŠ¶å›¾åˆ›å»ºå®Œæˆ")
    
    # 4. å †å æŸ±çŠ¶å›¾
    print(f"\n   åˆ›å»ºå †å æŸ±çŠ¶å›¾...")
    plt.figure(figsize=(12, 6))
    df_sales.plot(kind='bar', stacked=True, width=0.8)
    plt.title('å­£åº¦é”€å”®å †å å›¾', fontsize=16, fontweight='bold')
    plt.xlabel('äº§å“', fontsize=12)
    plt.ylabel('é”€å”®é¢', fontsize=12)
    plt.xticks(rotation=0)
    plt.legend(title='å­£åº¦')
    plt.tight_layout()
    plt.show()
    print(f"   âœ… å †å æŸ±çŠ¶å›¾åˆ›å»ºå®Œæˆ")
    
    # 5. æ•£ç‚¹å›¾çŸ©é˜µ
    print(f"\n   åˆ›å»ºæ•£ç‚¹å›¾çŸ©é˜µ...")
    scatter_data = pd.DataFrame(np.random.randn(100, 4), 
                                 columns=['å˜é‡A', 'å˜é‡B', 'å˜é‡C', 'å˜é‡D'])
    
    pd.plotting.scatter_matrix(scatter_data, figsize=(12, 12), 
                               alpha=0.6, diagonal='hist')
    plt.suptitle('å˜é‡å…³ç³»æ•£ç‚¹å›¾çŸ©é˜µ', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    print(f"   âœ… æ•£ç‚¹å›¾çŸ©é˜µåˆ›å»ºå®Œæˆ")
    
    # 6. ç®±çº¿å›¾
    print(f"\n   åˆ›å»ºç®±çº¿å›¾...")
    plt.figure(figsize=(10, 6))
    scatter_data.plot(kind='box')
    plt.title('å˜é‡åˆ†å¸ƒç®±çº¿å›¾', fontsize=16, fontweight='bold')
    plt.ylabel('æ•°å€¼', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f"   âœ… ç®±çº¿å›¾åˆ›å»ºå®Œæˆ")

demonstrate_dataframe_plotting()

# 2. LangChainæ•°æ®æ¢ç´¢å¯è§†åŒ–
print(f"\nğŸ¤– 2. LangChainæ•°æ®æ¢ç´¢å¯è§†åŒ–:")

@dataclass
class DataExplorer:
    """æ•°æ®æ¢ç´¢å™¨ - æ¨¡æ‹ŸLangChainçš„æ•°æ®æ¢ç´¢å¯è§†åŒ–"""
    
    def __init__(self):
        self.data = None
    
    def create_conversation_data(self, num_records: int = 200):
        """åˆ›å»ºå¯¹è¯æ•°æ®"""
        np.random.seed(42)
        
        dates = pd.date_range('2023-01-01', periods=num_records, freq='H')
        
        data = {
            'timestamp': dates,
            'user_id': [f'user_{i:03d}' for i in np.random.randint(1, 51, num_records)],
            'model': np.random.choice(['gpt-3.5-turbo', 'gpt-4', 'claude-2'], num_records),
            'category': np.random.choice(['æŠ€æœ¯', 'åˆ›æ„', 'åˆ†æ', 'é€šç”¨'], num_records),
            'tokens_used': np.random.randint(10, 150, num_records),
            'response_time_ms': np.random.randint(300, 3000, num_records),
            'satisfaction_score': np.random.uniform(3.0, 5.0, num_records),
            'cost_usd': np.random.uniform(0.01, 0.20, num_records),
            'success': np.random.choice([True, False], num_records, p=[0.9, 0.1])
        }
        
        self.data = pd.DataFrame(data)
        self.data['hour'] = self.data['timestamp'].dt.hour
        self.data['date'] = self.data['timestamp'].dt.date
    
    def quick_exploration(self):
        """å¿«é€Ÿæ•°æ®æ¢ç´¢"""
        if self.data is None:
            print("   è¯·å…ˆåˆ›å»ºæ•°æ®")
            return
        
        print(f"   å¿«é€Ÿæ•°æ®æ¢ç´¢å¯è§†åŒ–:")
        
        # 1. æ—¶é—´åºåˆ—è¶‹åŠ¿
        print(f"\n   1. æ—¶é—´åºåˆ—è¶‹åŠ¿...")
        daily_stats = self.data.groupby('date').agg({
            'tokens_used': 'sum',
            'cost_usd': 'sum',
            'satisfaction_score': 'mean'
        })
        
        fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
        
        daily_stats['tokens_used'].plot(ax=ax1, linewidth=2, color='blue')
        ax1.set_title('æ¯æ—¥Tokenä½¿ç”¨é‡è¶‹åŠ¿', fontweight='bold')
        ax1.set_ylabel('Tokenæ•°')
        ax1.grid(True, alpha=0.3)
        
        daily_stats['cost_usd'].plot(ax=ax2, linewidth=2, color='red')
        ax2.set_title('æ¯æ—¥æˆæœ¬è¶‹åŠ¿', fontweight='bold')
        ax2.set_ylabel('æˆæœ¬(USD)')
        ax2.grid(True, alpha=0.3)
        
        daily_stats['satisfaction_score'].plot(ax=ax3, linewidth=2, color='green')
        ax3.set_title('æ¯æ—¥å¹³å‡æ»¡æ„åº¦è¶‹åŠ¿', fontweight='bold')
        ax3.set_ylabel('æ»¡æ„åº¦')
        ax3.set_xlabel('æ—¥æœŸ')
        ax3.grid(True, alpha=0.3)
        ax3.set_ylim(3, 5)
        
        plt.tight_layout()
        plt.show()
        
        # 2. åˆ†å¸ƒåˆ†æ
        print(f"\n   2. åˆ†å¸ƒåˆ†æ...")
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        
        self.data['tokens_used'].plot(kind='hist', bins=30, ax=ax1, alpha=0.7, color='skyblue')
        ax1.set_title('Tokenä½¿ç”¨åˆ†å¸ƒ', fontweight='bold')
        ax1.set_xlabel('Tokenæ•°')
        
        self.data['response_time_ms'].plot(kind='hist', bins=30, ax=ax2, alpha=0.7, color='lightcoral')
        ax2.set_title('å“åº”æ—¶é—´åˆ†å¸ƒ', fontweight='bold')
        ax2.set_xlabel('å“åº”æ—¶é—´(ms)')
        
        self.data['satisfaction_score'].plot(kind='hist', bins=30, ax=ax3, alpha=0.7, color='lightgreen')
        ax3.set_title('æ»¡æ„åº¦åˆ†å¸ƒ', fontweight='bold')
        ax3.set_xlabel('æ»¡æ„åº¦')
        
        self.data['cost_usd'].plot(kind='hist', bins=30, ax=ax4, alpha=0.7, color='gold')
        ax4.set_title('æˆæœ¬åˆ†å¸ƒ', fontweight='bold')
        ax4.set_xlabel('æˆæœ¬(USD)')
        
        plt.tight_layout()
        plt.show()
        
        # 3. åˆ†ç±»åˆ†æ
        print(f"\n   3. åˆ†ç±»åˆ†æ...")
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        
        # æ¨¡å‹ä½¿ç”¨é¢‘æ¬¡
        model_counts = self.data['model'].value_counts()
        model_counts.plot(kind='bar', ax=ax1, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
        ax1.set_title('æ¨¡å‹ä½¿ç”¨é¢‘æ¬¡', fontweight='bold')
        ax1.set_ylabel('ä½¿ç”¨æ¬¡æ•°')
        
        # ç±»åˆ«ä½¿ç”¨é¢‘æ¬¡
        category_counts = self.data['category'].value_counts()
        category_counts.plot(kind='bar', ax=ax2, color=['#96CEB4', '#FFEAA7', '#DDA0DD', '#98D8C8'])
        ax2.set_title('ç±»åˆ«ä½¿ç”¨é¢‘æ¬¡', fontweight='bold')
        ax2.set_ylabel('ä½¿ç”¨æ¬¡æ•°')
        
        # æˆåŠŸç‡é¥¼å›¾
        success_counts = self.data['success'].value_counts()
        success_counts.plot(kind='pie', ax=ax3, labels=['æˆåŠŸ', 'å¤±è´¥'], 
                           autopct='%1.1f%%', colors=['#4CAF50', '#F44336'])
        ax3.set_title('æˆåŠŸç‡', fontweight='bold')
        
        # å°æ—¶åˆ†å¸ƒ
        hourly_counts = self.data['hour'].value_counts().sort_index()
        hourly_counts.plot(kind='bar', ax=ax4, color='orange')
        ax4.set_title('å°æ—¶åˆ†å¸ƒ', fontweight='bold')
        ax4.set_xlabel('å°æ—¶')
        ax4.set_ylabel('å¯¹è¯æ•°')
        
        plt.tight_layout()
        plt.show()
        
        # 4. å…³ç³»åˆ†æ
        print(f"\n   4. å…³ç³»åˆ†æ...")
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        
        # Token vs å“åº”æ—¶é—´
        self.data.plot(kind='scatter', x='tokens_used', y='response_time_ms', 
                      alpha=0.6, ax=ax1, color='blue')
        ax1.set_title('Token vs å“åº”æ—¶é—´', fontweight='bold')
        ax1.set_xlabel('Tokenæ•°')
        ax1.set_ylabel('å“åº”æ—¶é—´(ms)')
        
        # Token vs æˆæœ¬
        self.data.plot(kind='scatter', x='tokens_used', y='cost_usd', 
                      alpha=0.6, ax=ax2, color='red')
        ax2.set_title('Token vs æˆæœ¬', fontweight='bold')
        ax2.set_xlabel('Tokenæ•°')
        ax2.set_ylabel('æˆæœ¬(USD)')
        
        # æ»¡æ„åº¦ vs å“åº”æ—¶é—´
        self.data.plot(kind='scatter', x='satisfaction_score', y='response_time_ms', 
                      alpha=0.6, ax=ax3, color='green')
        ax3.set_title('æ»¡æ„åº¦ vs å“åº”æ—¶é—´', fontweight='bold')
        ax3.set_xlabel('æ»¡æ„åº¦')
        ax3.set_ylabel('å“åº”æ—¶é—´(ms)')
        
        # æ»¡æ„åº¦ vs æˆæœ¬
        self.data.plot(kind='scatter', x='satisfaction_score', y='cost_usd', 
                      alpha=0.6, ax=ax4, color='purple')
        ax4.set_title('æ»¡æ„åº¦ vs æˆæœ¬', fontweight='bold')
        ax4.set_xlabel('æ»¡æ„åº¦')
        ax4.set_ylabel('æˆæœ¬(USD)')
        
        plt.tight_layout()
        plt.show()
    
    def create_summary_report(self):
        """åˆ›å»ºæ±‡æ€»æŠ¥å‘Š"""
        if self.data is None:
            print("   è¯·å…ˆåˆ›å»ºæ•°æ®")
            return
        
        # è®¡ç®—å…³é”®æŒ‡æ ‡
        summary = {
            'æ€»å¯¹è¯æ•°': len(self.data),
            'æ´»è·ƒç”¨æˆ·æ•°': self.data['user_id'].nunique(),
            'å¹³å‡Tokenä½¿ç”¨': self.data['tokens_used'].mean(),
            'å¹³å‡å“åº”æ—¶é—´': self.data['response_time_ms'].mean(),
            'å¹³å‡æ»¡æ„åº¦': self.data['satisfaction_score'].mean(),
            'æ€»æˆæœ¬': self.data['cost_usd'].sum(),
            'æˆåŠŸç‡': self.data['success'].mean()
        }
        
        # åˆ›å»ºæ±‡æ€»å›¾è¡¨
        fig, axes = plt.subplots(2, 3, figsize=(18, 10))
        axes = axes.flatten()
        
        # 1. å…³é”®æŒ‡æ ‡æŸ±çŠ¶å›¾
        metrics = ['å¹³å‡Tokenä½¿ç”¨', 'å¹³å‡å“åº”æ—¶é—´', 'å¹³å‡æ»¡æ„åº¦', 'æ€»æˆæœ¬', 'æˆåŠŸç‡']
        values = [summary['å¹³å‡Tokenä½¿ç”¨'], summary['å¹³å‡å“åº”æ—¶é—´']/100, 
                 summary['å¹³å‡æ»¡æ„åº¦']*20, summary['æ€»æˆæœ¬']*100, summary['æˆåŠŸç‡']*100]
        
        axes[0].bar(metrics, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'orange'])
        axes[0].set_title('å…³é”®æŒ‡æ ‡', fontweight='bold')
        axes[0].tick_params(axis='x', rotation=45)
        
        # 2. æ¨¡å‹æ€§èƒ½å¯¹æ¯”
        model_perf = self.data.groupby('model').agg({
            'satisfaction_score': 'mean',
            'response_time_ms': 'mean',
            'cost_usd': 'mean'
        })
        
        model_perf.plot(kind='bar', ax=axes[1])
        axes[1].set_title('æ¨¡å‹æ€§èƒ½å¯¹æ¯”', fontweight='bold')
        axes[1].tick_params(axis='x', rotation=45)
        
        # 3. æ—¶é—´åˆ†å¸ƒ
        hourly_data = self.data.groupby('hour').size()
        hourly_data.plot(kind='line', ax=axes[2], marker='o')
        axes[2].set_title('24å°æ—¶åˆ†å¸ƒ', fontweight='bold')
        axes[2].set_xlabel('å°æ—¶')
        axes[2].set_ylabel('å¯¹è¯æ•°')
        
        # 4. ç±»åˆ«åˆ†å¸ƒ
        category_data = self.data.groupby('category').size()
        category_data.plot(kind='pie', ax=axes[3], autopct='%1.1f%%')
        axes[3].set_title('ç±»åˆ«åˆ†å¸ƒ', fontweight='bold')
        
        # 5. æ»¡æ„åº¦åˆ†å¸ƒ
        self.data['satisfaction_score'].plot(kind='hist', bins=20, ax=axes[4], alpha=0.7)
        axes[4].set_title('æ»¡æ„åº¦åˆ†å¸ƒ', fontweight='bold')
        axes[4].set_xlabel('æ»¡æ„åº¦')
        
        # 6. æˆæœ¬è¶‹åŠ¿
        daily_cost = self.data.groupby('date')['cost_usd'].sum()
        daily_cost.plot(kind='line', ax=axes[5], color='red')
        axes[5].set_title('æ¯æ—¥æˆæœ¬è¶‹åŠ¿', fontweight='bold')
        axes[5].set_xlabel('æ—¥æœŸ')
        axes[5].set_ylabel('æˆæœ¬(USD)')
        
        plt.tight_layout()
        plt.show()
        
        return summary

def demonstrate_data_explorer():
    """æ¼”ç¤ºæ•°æ®æ¢ç´¢å™¨"""
    explorer = DataExplorer()
    
    # åˆ›å»ºå¯¹è¯æ•°æ®
    print(f"   æ•°æ®æ¢ç´¢å¯è§†åŒ–æ¼”ç¤º:")
    explorer.create_conversation_data(200)
    print(f"   åˆ›å»ºäº† {len(explorer.data)} æ¡å¯¹è¯æ•°æ®")
    
    # å¿«é€Ÿæ¢ç´¢
    print(f"\n   æ‰§è¡Œå¿«é€Ÿæ•°æ®æ¢ç´¢...")
    explorer.quick_exploration()
    print(f"   âœ… å¿«é€Ÿæ•°æ®æ¢ç´¢å®Œæˆ")
    
    # æ±‡æ€»æŠ¥å‘Š
    print(f"\n   åˆ›å»ºæ±‡æ€»æŠ¥å‘Š...")
    summary = explorer.create_summary_report()
    print(f"   âœ… æ±‡æ€»æŠ¥å‘Šåˆ›å»ºå®Œæˆ")
    
    # æ˜¾ç¤ºå…³é”®æŒ‡æ ‡
    print(f"\n   ğŸ“Š æ±‡æ€»æŠ¥å‘Šå…³é”®æŒ‡æ ‡:")
    for key, value in summary.items():
        if isinstance(value, float):
            print(f"   {key}: {value:.2f}")
        else:
            print(f"   {key}: {value}")

demonstrate_data_explorer()

print(f"\nâœ… Pandaså†…ç½®å¯è§†åŒ–å®Œæˆ")
print(f"ğŸ¯ å­¦ä¹ ç›®æ ‡è¾¾æˆ:")
print(f"   âœ“ æŒæ¡Pandas plot()æ–¹æ³•çš„ä½¿ç”¨")
print(f"   âœ“ ç†è§£ä¸åŒå›¾è¡¨ç±»å‹çš„å¿«é€Ÿåˆ›å»º")
print(f"   âœ“ ç†Ÿç»ƒä½¿ç”¨Pandaså¯è§†åŒ–çš„å‚æ•°å®šåˆ¶")
print(f"   âœ“ èƒ½å¤Ÿç»“åˆæ•°æ®å¤„ç†å’Œå¯è§†åŒ–")
````
