### 遍历

#### 基本遍历

```python
text = "Python"

# 直接迭代（最常用）
for char in text:
    print(char, end=" ")  # P y t h o n

print()

# 索引遍历
for i in range(len(text)):
    print(text[i], end=" ")  # P y t h o n
```

#### 带索引遍历

```python
text = "Hello"

# enumerate（推荐）
for i, char in enumerate(text):
    print(f"索引{i}: '{char}'")

# range + 索引
for i in range(len(text)):
    char = text[i]
    print(f"位置{i}: {char}")
```

#### 反向遍历

```python
text = "Python"

# 切片反转
for char in text[::-1]:
    print(char, end=" ")  # n o h t y P

print()

# reversed函数
for char in reversed(text):
    print(char, end=" ")  # n o h t y P

print()

# 负索引
for i in range(len(text)-1, -1, -1):
    print(text[i], end=" ")  # n o h t y P
```

#### 步长遍历

```python
text = "Python编程"

# 方式8: 切片步长
# 每隔一个字符
for char in text[::2]:
    print(char, end=" ")  # P t o 编

print()

# 从第二个字符开始，每隔两个
for char in text[1::2]:
    print(char, end=" ")  # y h n程

print()

# 反向步长
for char in text[::-2]:
    print(char, end=" ")  # 程 n h y
```

### 格式化

#### %

```python
name = "小明"
age = 20
score = 95.5

# 方式1：%
old_format = "姓名：%s，年龄：%d，成绩：%.1f" % (name, age, score)
print(old_format)

# > 姓名：小明，年龄：20，成绩：95.5
```

#### format

```python
format_style = "姓名：{}，年龄：{}，成绩：{:.1f}".format(name, age, score)
print(format_style)

# > 姓名：小明，年龄：20，成绩：95.5
```

#### f-string

```python
f_string = f"姓名：{name}，年龄：{age}，成绩：{score:.1f}"
print(f_string)

# > 姓名：小明，年龄：20，成绩：95.5
```

| 符  号 | 描述                                 |
| :----- | :----------------------------------- |
| %c     | 格式化字符及其ASCII码                |
| %s     | 格式化字符串                         |
| %d     | 格式化整数                           |
| %u     | 格式化无符号整型                     |
| %o     | 格式化无符号八进制数                 |
| %x     | 格式化无符号十六进制数               |
| %X     | 格式化无符号十六进制数（大写）       |
| %f     | 格式化浮点数字，可指定小数点后的精度 |
| %e     | 用科学计数法格式化浮点数             |
| %E     | 作用同%e，用科学计数法格式化浮点数   |
| %g     | %f和%e的简写                         |
| %G     | %f 和 %E 的简写                      |
| %p     | 用十六进制数格式化变量的地址         |
