```python
fruits = ['apple', 'banana', 'orange']

# 不使用enumerate（麻烦）
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")

# 使用enumerate（推荐）
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

### 指定起始索引

```python
fruits = ['apple', 'banana', 'orange']

# 默认从0开始
for i, fruit in enumerate(fruits):
    print(f"{i}-{fruit}", end=" ")  # 0, 1, 2

# 从1开始
print()
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}-{fruit}", end=" ")  # 1., 2., 3.

# 从10开始
print()
for i, fruit in enumerate(fruits, start=10):
    print(f"{i}-{fruit}", end=" ")  # #10, #11, #12
```
