### 修改

```python
person = {'name': 'Bill', 'age': 10 }
person['age'] = 30
print(person)

del person['age']
print(person)
```

### 方法

| 方法 / 表达式                      | 方法描述                                      |
| ---------------------------------- | --------------------------------------------- |
| dict.clear()                       | 删除字典内所有元素                            |
| dict.copy()                        | 返回字典的浅复制                              |
| dict.fromkeys()                    | 以序列 seq 元素为键，val 为初始值创建新字典   |
| dict.get(key, default=None)        | 返回指定键的值，键不存在则返回默认值          |
| key in dict                        | 判断键是否在字典中，存在返回 True，否则 False |
| dict.items()                       | 以列表返回字典的视图对象                      |
| dict.keys()                        | 返回字典键的视图对象                          |
| dict.setdefault(key, default=None) | 类似 get ()，键不存在则添加键并设为默认值     |
| dict.update(dict2)                 | 将 dict2 的键值对更新到 dict 中               |
| dict.values()                      | 返回字典值的视图对象                          |
| dict.pop(key[,default])            | 删除指定键对应值并返回，可指定默认值          |
| dict.popitem()                     | 返回并删除字典最后一对键值对                  |
