### 基础使用

```python
# 打印字符串
print("Hello World")  # Hello World

# 打印多个值
print("姓名:", "张三", "年龄:", 25)  # 姓名: 张三 年龄: 25

# 打印变量
name = "李四"
age = 30
print(name, age)  # 李四 30
```

### sep & end (分隔符和结束符)

```python
# sep 分隔符
# sep: str | None = " "
print("Python", "Java", "C++", sep=", ") # 逗号分隔：Python, Java, C++

# end 结束符
# end: str | None = "\n"
print("Hello", end=" ")     # 不换行，加空格
print("World")              # Hello World
```

## 实用技巧

### 进度条

```python
import time

print("下载进度:")
for i in range(1, 21):
    bar = "█" * i + "░" * (20 - i)
    percent = i * 5
    print(f"\r[{bar}] {percent}%", end="", flush=True)
    time.sleep(0.1)
print()  # 换行
```

### 彩色输出

```python
# 方法1：ANSI转义码
print("\033[31m红色文字\033[0m")      # 红色
print("\033[32m绿色文字\033[0m")      # 绿色
print("\033[33m黄色文字\033[0m")      # 黄色
print("\033[34m蓝色文字\033[0m")      # 蓝色
print("\033[1m粗体文字\033[0m")       # 粗体
print("\033[4m下划线文字\033[0m")     # 下划线

# 方法2：使用colorama库（跨平台）
from colorama import Fore, Back, Style, init
init(autoreset=True)  # 自动重置颜色

print(Fore.RED + "红色文字")
print(Fore.GREEN + "绿色文字")
print(Fore.BLUE + Back.YELLOW + "蓝字黄底")
print(Style.BRIGHT + "亮色文字")
```

### 表格输出

```python
# 简单表格
data = [
    ["张三", 25, "北京"],
    ["李四", 30, "上海"],
    ["王五", 35, "广州"]
]

print("姓名    年龄  城市")
print("-" * 20)
for row in data:
    print(f"{row[0]:<8}{row[1]:<6}{row[2]:<6}")

# 使用tabulate库（需安装）
from tabulate import tabulate
headers = ["姓名", "年龄", "城市"]
print(tabulate(data, headers=headers, tablefmt="grid"))
```

### 代码块耗时

```python
import time

start = time.time()
# 模拟耗时操作
time.sleep(1)
end = time.time()

print(f"耗时: {end - start:.2f}秒")  # 耗时: 1.00秒
print(f"耗时: {end - start:.3f}秒")  # 耗时: 1.000秒
```
