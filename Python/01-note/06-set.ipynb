- 无序
- 不支持索引访问
- 不支持切片操作
- 不重复
- 支持交集、并集、差集操作

### 删除

```python
set_data = {1, 2, 3}

# 移除 - 方式1
set_data.remove(3)
print(set_data)

# 移除 - 方式2
set_data.discard(2)
print(set_data)

# 不支持
# del set[2]

# set.remove(5) # 不存在，会报错
set_data.discard(5) # 不存在，不会报错 

x = set_data.pop()     # 随机移除并返回一个元素
print(x, set_data)     
```

### 元素去重

本质是计算哈希表

| 类型                 | 是否可哈希           | 示例                   | 可放入集合           |
| :------------------- | :------------------- | :--------------------- | :------------------- |
| **int, float, bool** | ✅ 是                 | `1`, `3.14`, `True`    | ✓                    |
| **str**              | ✅ 是                 | `"hello"`              | ✓                    |
| **tuple**            | ✅ 是（元素全可哈希） | `(1, 2)`, `("a", "b")` | ✓                    |
| **frozenset**        | ✅ 是                 | `frozenset([1, 2])`    | ✓                    |
| **list**             | ❌ 否                 | `[1, 2]`               | ✗                    |
| **dict**             | ❌ 否                 | `{"a": 1}`             | ✗                    |
| **set**              | ❌ 否                 | `{1, 2}`               | ✗                    |
| **自定义类**         | 取决于实现           | `Person("Alice")`      | 如果实现了`__hash__` |

```python
# 1. 基础数据类型去重
numbers = [1, 2, 2, 3, 3, 3]
print(set(numbers))  # {1, 2, 3}

strings = ["a", "b", "a", "c", "b"]
print(set(strings))  # {'a', 'b', 'c'}

# 2. 混合类型去重
mixed = [1, 1.0, True, "1", (1,), frozenset([1, 2])]  # 注意：1 == 1.0 == True
print(set(mixed))  
```

### 集合运算

```python
A = {1, 2, 3}
B = {2, 3, 4}

print(A | B)   # 并集
print(A & B)   # 交集
print(A - B)   # 差集
print(A ^ B)   # 对称差
```

### 成员检测

```python
set_data = {1, 2, 3, 4}
print(2 in set_data)    # True
print(5 not in set_data) # True
```

### 集合比较

```python
A = {1, 2}
B = {1, 2, 3}
C = {1, 2}

print(A <= B)   # 子集: True (A⊆B)
print(B >= A)   # 超集: True (B⊇A)
print(A < B)    # 真子集: True (A⊂B)
print(A == C)   # 相等: True
```

### 集合推导式

```python
# 创建0-9的偶数集合
even_set = {x for x in range(10) if x % 2 == 0}
print(even_set)  # {0, 2, 4, 6, 8}
```
