- match 值: 开头
- case 模式: 匹配各种模式
- case _: 默认匹配
- 按顺序匹配，匹配到就停止


匹配类型
- 字面值
- 类型
- 序列
- 映射
- 类
- 或模式
- 守卫
- 通配符
- as 绑定

### 匹配值

```python
x = 2
match x:
    case 1:
        print("一")
    case 2:
        print("二")
    case 3:
        print("三")
    case _:
        print("其他")
```

支持混合类型

```python
# 直接匹配值
value = 3.14

match value:
    case 1:          # 整数
        print(1)
    case "hello":    # 字符串
        print("hello")
    case 3.14:       # 浮点数
        print(3.14)
    case True:       # 布尔值
        print(True)
    case None:       # None值
        print(None)
```

### 类型匹配

```python
def check_type(val):
    match val:
        case int():
            return "整数"
        case str():      # 字符串
            return "字符串"
        case list():     # 列表
            return "列表"
        case dict():     # 字典
            return "字典"
        case float():    # 浮点数
            return "浮点数"

print(check_type(10))    # 整数
print(check_type("hi"))  # 字符串
```

## 容器类型匹配

### 列表匹配

```python
data = [2, 2, 3, 4, 5]
match data:
    case []:
        print("空列表")
    case [x]:
        print(f"单元素: {x}")
    case [x, y]:
        print(f"两个元素: {x}, {y}")
    case [1, 2, *tail]:
        print(f"以1,2开头: {tail}")
    case [x, y, *rest]: 
        print(f"多个元素: {x}, {y}, {rest}")
    case _:
        print("多个元素")
```

### 元组匹配

```python
point = (2, 3)
match point:
    case (0, 0):
        print("原点")
    case (x, 0):
        print(f"x轴上: {x}")
    case (0, y):
        print(f"y轴上: {y}")
    case (x, y):
        print(f"点({x}, {y})")
```

### 字典匹配

部分匹配

```python
data = {"name": "张三", "age": 25}

match data:
    # case {}: # 有时会输出"空字典"（bug！）
    #     print("空字典")

    # 匹配单个 key
    case {"name": n}:          
        print(f"姓名: {n}")

    # 匹配多个 key
    case {"name": n, "age": a}: # 两个键都有
        print(f"{n}, {a}岁")

    # 匹配单组 key-value
    case {"name": "张三"}:           
        print(r"{name: 张三}")

    # 匹配多组 key-value
    case {"name": "张三", "age": 25}:
        print("张三, 25岁")

    # 带守卫条件的匹配
    case {"name": n, "age": a} if a >= 18:  # 两个键都有且年龄>=18
        print(f"{n}, {a}岁")
```

### 带条件的匹配

```python
num = 15
match num:
    case num if num < 0:
        print("负数")
    case num if num < 10:
        print("个位数")
    case num if num < 100:
        print("两位数")
    case _:
        print("大数")
```

### 或匹配（|）

```python
# 多个值匹配
status = 404
match status:
    case 200 | 201:
        print("成功")
    case 400 | 404:
        print("客户端错误")
    case 500:
        print("服务器错误")
# 输出: 客户端错误
```
