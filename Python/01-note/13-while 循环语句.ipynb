### 条件判断

```python
count = 0
while count <3:
    print(f"计数: {count}")
    count += 1
```

### 无限循环

```python
count = 0
while True: 
    print(f"计数: {count}")
    count += 1
    if count == 3: break
```

### break 和 continue

```python
count = 0
while True: 
    count += 1
    if count == 6: break        # 累计到指定数值，退出循环
    if count % 2 == 0: continue # 如果是偶数，执行下一次循环
    print(count, end=" ")
```

### while-else

```python
count = 0

while count < 2: 
   count += 1
   print(count, end=" ")
else:
    print("End") 
```
