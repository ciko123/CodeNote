# 简单练习5：猜数字游戏

## 题目描述
创建一个互动的猜数字游戏，程序随机生成一个数字，玩家需要在有限次数内猜中。

## 具体要求
1. **游戏设置**：初始化游戏参数
   - 随机生成1-100之间的整数作为目标数字
   - 设置最大猜测次数为7次
   - 显示游戏规则和欢迎信息

2. **游戏流程**：实现完整的游戏逻辑
   - 每次提示玩家输入猜测的数字
   - 验证输入是否为有效数字（1-100范围内）
   - 比较猜测数字与目标数字，给出提示
   - 显示剩余猜测次数
   - 处理各种异常情况（非数字输入、超出范围等）

3. **反馈机制**：提供清晰的游戏反馈
   - 猜对了：显示祝贺信息和所用次数
   - 猜小了：提示"太小了! 再大一点"
   - 猜大了：提示"太大了! 再小一点"
   - 次数用完：显示游戏结束和正确答案

4. **统计功能**：记录和分析游戏数据
   - 模拟多次游戏结果（3, 5, 2, 7, 4, 6, 3, 5次）
   - 计算平均猜测次数
   - 显示最佳和最差记录
   - 计算5次内猜中的成功率

## 游戏规则
- 程序随机生成1-100的数字
- 玩家有7次猜测机会
- 每次猜测后会提示数字是太大还是太小
- 在7次内猜中即为胜利

## 预期输出
```
欢迎来到猜数字游戏!
规则: 在7次内猜中1-100的随机数字

第1次猜测 (1-100): 50

第2次猜测 (1-100): 25

=== 游戏统计 ===

最佳记录: 2次
最差记录: 7次

```

## 异常处理示例
```
第1次猜测 (1-100): abc
请输入有效的数字!

第1次猜测 (1-100): 150
请输入1-100之间的数字!
```

## 实现提示
- 使用random.randint生成随机数
- 使用while循环控制游戏流程
- 使用try-except处理输入异常
- 使用int()转换输入为数字
- 使用列表推导式统计成功率
- 使用内置函数sum、min、max进行统计

## 答案

````python
# 猜数字游戏 - 涵盖随机数、循环、条件判断、输入输出、异常处理

import random

def guess_number_game():
    print("=== 猜数字游戏 ===")
    print("我想了一个1-100的数字，你来猜!")
    
    # 生成随机数
    target = random.randint(1, 100)
    attempts = 0
    max_attempts = 7
    
    while attempts < max_attempts:
        try:
            # 获取用户输入
            guess = int(input(f"\n第{attempts + 1}次猜测 (1-100): "))
            attempts += 1
            
            # 验证输入范围
            if guess < 1 or guess > 100:
                print("请输入1-100之间的数字!")
                continue
            
            # 判断猜测结果
            if guess == target:
                print(f"🎉 恭喜! 你用{attempts}次猜中了数字{target}!")
                return
            elif guess < target:
                print("太小了! 再大一点")
            else:
                print("太大了! 再小一点")
                
            # 显示剩余机会
            remaining = max_attempts - attempts
            if remaining > 0:
                print(f"还有{remaining}次机会")
                
        except ValueError:
            print("请输入有效的数字!")
            attempts -= 1  # 不计算无效输入
    
    # 游戏结束
    print(f"\n游戏结束! 正确答案是{target}")
    if attempts >= max_attempts:
        print("下次再加油吧!")

def game_statistics():
    """显示游戏统计"""
    print("\n=== 游戏统计 ===")
    
    # 模拟多次游戏结果
    game_results = [3, 5, 2, 7, 4, 6, 3, 5]  # 猜测次数
    
    if game_results:
        avg_attempts = sum(game_results) / len(game_results)
        best_game = min(game_results)
        worst_game = max(game_results)
        
        print(f"游戏次数: {len(game_results)}")
        print(f"平均猜测次数: {avg_attempts:.1f}")
        print(f"最佳记录: {best_game}次")
        print(f"最差记录: {worst_game}次")
        
        # 成功率统计
        success_count = sum(1 for x in game_results if x <= 5)
        success_rate = (success_count / len(game_results)) * 100
        print(f"5次内猜中成功率: {success_rate:.1f}%")

# 运行游戏
print("欢迎来到猜数字游戏!")
print("规则: 在7次内猜中1-100的随机数字")
print()

# 开始游戏
guess_number_game()

# 显示统计
game_statistics()

# 知识点总结：
# 1. 随机数生成 (random.randint)
# 2. while循环控制
# 3. 条件判断 (if-elif-else)
# 4. 异常处理 (try-except)
# 5. 输入验证和类型转换 (int, ValueError)
# 6. 列表操作和统计 (sum, min, max)
# 7. 列表推导式和计数
````
