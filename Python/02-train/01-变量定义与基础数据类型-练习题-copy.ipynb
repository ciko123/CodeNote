### ç¬¬8é¢˜ç­”æ¡ˆï¼šå®‰å…¨ç±»å‹è½¬æ¢å‡½æ•°

````python
# 1. ç¼–å†™å®‰å…¨è½¬æ¢å‡½æ•°
def safe_convert(value, target_type, default=None):
    """
    å®‰å…¨ç±»å‹è½¬æ¢å‡½æ•°
    
    å‚æ•°:
        value: å¾…è½¬æ¢çš„å€¼
        target_type: ç›®æ ‡ç±»å‹ (int, float, str, bool)
        default: è½¬æ¢å¤±è´¥æ—¶çš„é»˜è®¤å€¼
    
    è¿”å›:
        è½¬æ¢æˆåŠŸè¿”å›è½¬æ¢åçš„å€¼ï¼Œå¤±è´¥è¿”å›é»˜è®¤å€¼
    """
    try:
        if target_type == bool:
            # ç‰¹æ®Šå¤„ç†å¸ƒå°”ç±»å‹
            if isinstance(value, str):
                return value.lower() in ('true', '1', 'yes')
            return bool(value)
        else:
            return target_type(value)
    except (ValueError, TypeError):
        return default

# 2. æµ‹è¯•è½¬æ¢
print("=== å®‰å…¨ç±»å‹è½¬æ¢æµ‹è¯• ===")

test_cases = [
    ("123", int, 0),
    ("abc", int, -1),
    ("3.14", float, 0.0),
    ("True", bool, False),
    (None, str, "N/A")
]

for value, target_type, default in test_cases:
    result = safe_convert(value, target_type, default)
    print(f"è½¬æ¢ '{value}' ä¸º {target_type.__name__}: {result} (ç±»å‹: {type(result).__name__})")

# é¢å¤–æµ‹è¯•
print("\n=== é¢å¤–æµ‹è¯• ===")
print(f"'0' è½¬ bool: {safe_convert('0', bool, False)}")
print(f"'false' è½¬ bool: {safe_convert('false', bool, True)}")
print(f"'3.14.15' è½¬ float: {safe_convert('3.14.15', float, 0.0)} (æ— æ•ˆæ ¼å¼)")
````
