# 简单练习1：迷你计算器

## 题目描述
创建一个简单的命令行计算器程序，能够进行基本的数学运算。

## 具体要求
1. **输入功能**：程序需要获取用户输入的两个数字和一个运算符
   - 第一个数字：任意实数（整数或小数）
   - 运算符：+、-、*、/ 四种之一
   - 第二个数字：任意实数（整数或小数）

2. **运算功能**：根据运算符执行相应的数学运算
   - 加法（+）：计算两个数的和
   - 减法（-）：计算第一个数减去第二个数
   - 乘法（*）：计算两个数的积
   - 除法（/）：计算第一个数除以第二个数

3. **错误处理**：处理以下异常情况
   - 除数为0时显示错误信息
   - 输入无效运算符时显示错误信息
   - 输入非数字时显示错误信息

4. **输出格式**：显示完整的计算过程和结果
   - 格式：`数字1 运算符 数字2 = 结果`
   - 小数结果保留2位小数

## 示例运行
```
=== 迷你计算器 ===
请输入第一个数字: 10.5
请输入运算符 (+, -, *, /): *
请输入第二个数字: 2
10.5 * 2 = 21.00
```

## 错误处理示例
```
=== 迷你计算器 ===
请输入第一个数字: 10
请输入运算符 (+, -, *, /): /
请输入第二个数字: 0
错误：除数不能为0
```

## 实现提示
- 使用 `input()` 函数获取用户输入
- 使用 `float()` 将字符串转换为数字
- 使用 `if-elif-else` 语句判断运算符
- 使用函数封装计算器逻辑

## 答案

````python


# 1. 拿到用户的输出


# 2. 解析用户的输入是否符合规范
#  - 根据包含的运算符分割
#  - 分割得到的两个数值去空格
#  - 根据不同的类型，使用不同的法法得到值
#  - 除数不能为 0，为0需要提醒

# 只能包含一个运算符



def calculate():

    try:
        # 获取用户输入
        num1 = float(input("请输入第一个数字: "))
        op = input("请输入运算符 (+, -, *, /): ").strip()
        if(op not in ['+','-','*','/']):
            print("错误警告：请输入正确的运算符")
            return
        num2 = float(input("请输入第二个数字: "))
    except ValueError as e:
        print("错误警告：请输入正确的数字") 
    except Exception as e:
        print("未知错误：", e)

    if(op == '+'):
        result = num1 + num2
    elif(op == '-'):
        result = num1 - num2
    elif(op == '*'):
        result = num1 * num2
    elif(op == '/'):
        if(num2 == 0):
            print("错误警告：除数不能为0")
            return
        else:
            result = num1 / num2

    print(f"{num1} {op} {num2} = {result}")


calculate()
````

````python
# 迷你计算器 - 涵盖变量、输入输出、运算符、条件判断、函数

def calculator():
    print("=== 迷你计算器 ===")
    
    # 获取用户输入
    num1 = float(input("请输入第一个数字: "))
    op = input("请输入运算符 (+, -, *, /): ")
    num2 = float(input("请输入第二个数字: "))
    
    # 条件判断进行计算
    if op == "+":
        result = num1 + num2
        print(f"{num1} + {num2} = {result}")
    elif op == "-":
        result = num1 - num2
        print(f"{num1} - {num2} = {result}")
    elif op == "*":
        result = num1 * num2
        print(f"{num1} * {num2} = {result}")
    elif op == "/":
        if num2 != 0:
            result = num1 / num2
            print(f"{num1} / {num2} = {result}")
        else:
            print("错误：除数不能为0")
    else:
        print("错误：未知运算符")

# 运行计算器
calculator()

# 知识点总结：
# 1. 变量定义 (num1, num2, op, result)
# 2. 输入输出 (input, print)
# 3. 数据类型转换 (float)
# 4. 算术运算符 (+, -, *, /)
# 5. 条件判断 (if-elif-else)
# 6. 函数定义和调用
````
