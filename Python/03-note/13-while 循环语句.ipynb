### 条件判断

```python
count = 0
while count <3:
    print(f"计数: {count}")
    count += 1
```

### 无限循环

```python
count = 0
while True: 
    print(f"计数: {count}")
    count += 1
    if count == 3: break
```

### break 和 continue

```python
count = 0
while True: 
    print(f"计数: {count}")
    count += 1
    if count == 3: break
```

```python
count = 0
while count <= 5: 
    count += 1
    if count % 2 == 0: continue
    print(f"计数: {count}")
```

### while-else

```python
# 查找列表中的元素
numbers = [1, 3, 5]
target = 4
i = 0

while i < len(numbers):
    if numbers[i] == target:
        print(f"找到 {target}")
        break
    i += 1
else:
    print(f"未找到 {target}")  # 循环正常结束执行
```
