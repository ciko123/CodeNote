### 基础随机

```python
import random

# 随机浮点数 0.0 <= x < 1.0
print(random.random())  # 0.123456...

# 随机整数 [a, b]
print(random.randint(1, 10))  # 1-10的随机整数

# 随机范围 [a, b) 步长step
print(random.randrange(0, 100, 5))  # 0,5,10,...,95
```

### 序列随机

```python
import random

items = ['苹果', '香蕉', '橙子', '葡萄', '西瓜']

# 随机选择一个
print(random.choice(items))  # 随机一个水果

# 随机选择多个（可重复）
print(random.choices(items, k=3))  # ['苹果', '葡萄', '苹果']

# 随机选择多个（不重复）
print(random.sample(items, 3))  # ['香蕉', '西瓜', '橙子']

# 打乱顺序
random.shuffle(items)
print(items)  # 顺序被打乱
```

### 随机分布

```python
import random

# 均匀分布 [a, b]
print(random.uniform(1.5, 10.5))  # 1.5-10.5的随机浮点数

# 正态分布 (均值, 标准差)
print(random.normalvariate(100, 15))  # 均值100, 标准差15

# 指数分布 (lambda)
print(random.expovariate(1.0))  # lambda=1.0
```

### 随机范围

```python
import random

# 常见随机范围生成

# 0-1之间的随机小数
print(random.random())

# a-b之间的随机整数（包含a和b）
print(random.randint(0, 100))

# a-b之间的随机小数
print(random.uniform(0, 1))

# 从序列中随机选择
colors = ['红', '绿', '蓝']
print(random.choice(colors))

# 从可迭代对象中随机选择k个
print(random.sample(range(100), 5))

# 随机打乱顺序
cards = list(range(10))
random.shuffle(cards)
print(cards)
```
