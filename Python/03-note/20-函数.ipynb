```python
# *args: 
def sum_all(*args):
    """"""
    total = 0
    for num in args:
        total += num
    return total

print(sum_all(1, 2, 3))        # 6
print(sum_all(1, 2, 3, 4, 5))  # 15

# **kwargs: 
def print_info(**kwargs):
    """"""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, city="Beijing")
```

```python
def complex_func(a, b, *args, option=True, **kwargs):
    """
    
    1. 
    2. *args
    3. 
    4. **kwargs
    """
    print(f"a={a}, b={b}")
    print(f"args={args}")
    print(f"option={option}")
    print(f"kwargs={kwargs}")

complex_func(1, 2, 3, 4, 5, option=False, x=10, y=20)
```

### *args  **kwargs

```python
# *args 
def sum_numbers(*args):
    """"""
    print(f"args : {type(args)}")  # <class 'tuple'>
    print(f"args : {args}")        # 
    return sum(args)

print(sum_numbers(1, 2, 3))        # 6
print(sum_numbers(1, 2, 3, 4, 5))  # 15
```

```python
# **kwargs 
def print_info(**kwargs):
    """"""
    print(f"kwargs : {type(kwargs)}")  # <class 'dict'>
    print(f"kwargs : {kwargs}")        # 
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, city="Beijing")
# :
# name: Alice
# age: 25
# city: Beijing
```

```python
#   *args    **kwargs
def func(a, b, *args, option=True, **kwargs):
    """
    a, b: 
    *args: 
    option: 
    **kwargs: 
    """
    print(f"a={a}, b={b}")
    print(f"args={args}")
    print(f"option={option}")
    print(f"kwargs={kwargs}")

# 
func(1, 2, 3, 4, 5, option=False, x=10, y=20)
# :
# a=1, b=2
# args=(3, 4, 5)
# option=False
# kwargs={'x': 10, 'y': 20}
```

```python
def apply_function(func, value):
    """"""
    return func(value)

def double(x):
    return x * 2

print(apply_function(double, 5))      # 10
print(apply_function(lambda x: x*3, 5))  # 15
```

### 

```python
def create_multiplier(factor):
    """"""
    def multiplier(x):
        return x * factor
    return multiplier

double = create_multiplier(2)
triple = create_multiplier(3)

print(double(5))  # 10
print(triple(5))  # 15
```

```python
def counter(start=0):
    """"""
    count = start
    
    def increment():
        nonlocal count  # 
        count += 1
        return count
    
    return increment

c = counter(10)
print(c())  # 11
print(c())  # 12
print(c())  # 13
```
