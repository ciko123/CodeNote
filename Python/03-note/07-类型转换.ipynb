```python
# å®‰å…¨ç±»å‹è½¬æ¢
print(f"\nğŸ“ 5. å®‰å…¨ç±»å‹è½¬æ¢:")
def safe_int_convert(value, default=0):
    """å®‰å…¨è½¬æ¢ä¸ºæ•´æ•°"""
    try:
        return int(value)
    except (ValueError, TypeError):
        return default

test_values = ["123", "abc", "3.14", None, "456"]
for val in test_values:
    result = safe_int_convert(val, -1)
    print(f"   '{val}' -> å®‰å…¨è½¬æ¢ç»“æœ: {result}")
```

```python
# ç±»å‹è½¬æ¢çš„å®é™…åº”ç”¨
print(f"\nğŸ“ 6. ç±»å‹è½¬æ¢å®é™…åº”ç”¨:")
# æ¨¡æ‹Ÿç”¨æˆ·è¾“å…¥å¤„ç†
user_inputs = ["25", "175.5", "True", "invalid"]
processed_data = []

for input_val in user_inputs:
    processed = {
        'original': input_val,
        'as_int': safe_int_convert(input_val),
        'as_float': safe_int_convert(input_val) if input_val.replace('.', '').isdigit() else 0.0,
        'as_bool': bool(input_val) and input_val.lower() != 'false'
    }
    processed_data.append(processed)

for data in processed_data:
    print(f"   è¾“å…¥: {data['original']}")
    print(f"   -> æ•´æ•°: {data['as_int']}, æµ®ç‚¹æ•°: {data['as_float']}, å¸ƒå°”: {data['as_bool']}")
```
