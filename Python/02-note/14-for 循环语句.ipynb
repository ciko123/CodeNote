### 遍历字符串

标准写法

```python
for char in "Python":
    print(char, end=", ")  # P y t h o n
```

一行写法

```python
for char in "Hello": print(char, end=", ")
```

### break 和 continue

```python
# 控制循环流程
print("1. 遇到3时停止:")
for i in range(1, 6):
    if i == 3:
        break
    print(i, end=" ")  # 1 2

print("\n\n2. 跳过偶数:")
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i, end=" ")  # 1 3 5 7 9
```

### 遍历列表

```python
fruits = ["苹果", "香蕉", "橙子"]
for fruit in fruits:
    print(f"水果: {fruit}")
# 输出: 水果: 苹果 水果: 香蕉 水果: 橙子
```

### 遍历字典

```python
# 遍历字典的不同方式
student = {"name": "张三", "age": 20, "score": 95}

print("1. 遍历键:")
for key in student:
    print(f"键: {key}")

print("\n2. 遍历值:")
for value in student.values():
    print(f"值: {value}")

print("\n3. 遍历键值对:")
for key, value in student.items():
    print(f"{key}: {value}")
```

### range() 函数使用

```python
print("默认:")
for i in range(5):  # 0-4
    print(i, end=" ")

print("\n起始值:")
for i in range(2, 8):  # 2-7
    print(i, end=" ")

print("\n步长:")
for i in range(1, 10, 2):  # 1,3,5,7,9
    print(i, end=" ")

print("\n倒序:")
for i in range(10, 0, -1):  # 10到1
    print(i, end=" ")
```

### zip 同时遍历多个序列

```python
# 解压
names = ["张三", "李四", "王五"]
scores = [85, 92, 78]
pairs = list(zip(names, scores))  # [('Alice', 85), ('Bob', 92), ('Charlie', 78)]
```

```python
names = ["张三", "李四", "王五"]
scores = [85, 92, 78]
subjects = ["数学", "英语", "编程"]

print("学生成绩单:")
for name, score, subject in zip(names, scores, subjects):
    print(f"{name} 的 {subject} 成绩: {score}")
```

### enumerate

- 支持索引 index

```python
colors = ["红色", "绿色", "蓝色", "黄色", "紫色"]

print("带索引的遍历:")
for index, color in enumerate(colors):
    print(f"{index}: {color}")

print("\n从指定索引开始:")
for i, color in enumerate(colors, 2):
    print(f"{i}. {color}")
```

### 解包嵌套结构

```python
# 嵌套解包
points = [(1, 2), (3, 4)]
for x, y in points:
    print(f"坐标: ({x}, {y})")

# 嵌套列表解包
matrix = [[1, 2, 3], [4, 5, 6]]
for (a, b, c) in matrix:
    print(f"行: {a}, {b}, {c}")
```

### reversed() 反向迭代

```python
numbers = [1, 2, 3]
for num in reversed(numbers):
    print(num, end=" ")
```

### sorted() 排序迭代

数据类型

```python
list = [3, 2, 1, 5, 4]
for value in sorted(list):
    print(value, end=" ")
```

字符类型

```python
list = ['b', 'a', 'c']
for value in sorted(list):
    print(value, end=" ")
```

复杂类型

```python
# 按不同方式遍历
students = [
    ("张三", 85),
    ("李四", 92),
    ("王五", 78)
]

print("按姓名排序:")
for name, score in sorted(students):
    print(f"{name}: {score}")

print("\n按成绩降序:")
for name, score in sorted(students, key=lambda x: x[1], reverse=True):
    print(f"{name}: {score}")
```
