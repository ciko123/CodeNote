### if-elif-else

```python
temperature = 25

if temperature > 30:
    print("热")
elif temperature > 20:
    print("暖")
else:
    print("冷")
```

支持嵌套

```python
number = 7

if number > 0:
    print("这是一个正数", end="") # 不换行
    if number % 2 == 0:  # 嵌套if语句
        print("，而且是偶数")
    else:
        print("，而且是奇数")
elif number < 0:
    print("这是一个负数", end="")
    if number % 2 == 0:
        print("，而且是偶数")
    else:
        print("，而且是奇数")
else:
    print("这是零")
```

### match...case

```python
status=400
match status:
    case 400:
        print("Bad request")
    case 404:
        print("Not found")
    case _:
        print("Something's wrong with the internet")
```
