### 字符索引

```python
# 字符串访问
text = "0123456789"
print(text[0])          # 第一个字符
print(text[-1])         # 最后一个字符
print(text[2:])         # 脚本2之后的所有字符
print(text[0:6])        # 0开始，6结束，不包含6
print(text[0:-1])       # 输出第一个到倒数第二个的所有字符
print(text[2:8:2])      # 2开始，8结束，不包含8，步长为2
```

### 字符串格式化

#### %

```python
name = "小明"
age = 20
score = 95.5

# 方式1：%
old_format = "姓名：%s，年龄：%d，成绩：%.1f" % (name, age, score)
print(old_format)

# > 姓名：小明，年龄：20，成绩：95.5
```

#### format

```python
format_style = "姓名：{}，年龄：{}，成绩：{:.1f}".format(name, age, score)
print(format_style)

# > 姓名：小明，年龄：20，成绩：95.5
```

#### f-string

```python
f_string = f"姓名：{name}，年龄：{age}，成绩：{score:.1f}"
print(f_string)

# > 姓名：小明，年龄：20，成绩：95.5
```

| 符  号 | 描述                                 |
| :----- | :----------------------------------- |
| %c     | 格式化字符及其ASCII码                |
| %s     | 格式化字符串                         |
| %d     | 格式化整数                           |
| %u     | 格式化无符号整型                     |
| %o     | 格式化无符号八进制数                 |
| %x     | 格式化无符号十六进制数               |
| %X     | 格式化无符号十六进制数（大写）       |
| %f     | 格式化浮点数字，可指定小数点后的精度 |
| %e     | 用科学计数法格式化浮点数             |
| %E     | 作用同%e，用科学计数法格式化浮点数   |
| %g     | %f和%e的简写                         |
| %G     | %f 和 %E 的简写                      |
| %p     | 用十六进制数格式化变量的地址         |
